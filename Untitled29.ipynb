```python
# 📌 Importiere benötigte Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zeta, log
from scipy.fftpack import fft

# 📌 Schritt 1: Definiere S(T) = (1/pi) * Im(log ζ(1/2 + iT))
def S_T(T_values):
    """Berechnet S(T) für eine gegebene Menge an T-Werten."""
    S_values = []
    for T in T_values:
        s_t = (1 / np.pi) * np.imag(log(zeta(0.5 + 1j * T)))
        S_values.append(s_t)
    return np.array(S_values)

# 📌 Schritt 2: Wertebereich für T definieren
T_min, T_max = 10, 10000  # Bereich der Nullstellenhöhe
num_points = 1000         # Anzahl der Werte für T
T_values = np.linspace(T_min, T_max, num_points)

# 📌 Schritt 3: Berechne S(T)
S_values = S_T(T_values)

# 📌 Schritt 4: Plot von S(T)
plt.figure(figsize=(8, 5))
plt.plot(T_values, S_values, color='blue')
plt.xlabel("T")
plt.ylabel("S(T)")
plt.title("Numerische Berechnung von S(T)")
plt.grid()
plt.show()

# 📌 Schritt 5: Fourier-Analyse von S(T)
fft_values = fft(S_values)
freqs = np.fft.fftfreq(len(S_values), d=(T_max - T_min) / num_points)

# 📌 Schritt 6: Plot der Fourier-Analyse
plt.figure(figsize=(8, 5))
plt.plot(freqs[:len(freqs)//2], np.abs(fft_values[:len(fft_values)//2]), color="green")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Stärke")
plt.title("Fourier-Analyse von S(T)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
from scipy.fftpack import fft
from scipy.integrate import cumtrapz
from google.colab import drive

# 🔹 **1. Google Drive mounten und Daten laden**
drive.mount('/content/drive')
file_path = "/content/drive/MyDrive/zeros6.txt"

# Nullstellen einlesen
zeta_zeros = np.loadtxt(file_path)
N = len(zeta_zeros)
print(f"Anzahl geladener Nullstellen: {N}")

# 🔹 **2. Berechnung von S(T)**
def compute_S_T(zeros):
    T = zeros
    S_T = np.arange(1, len(T) + 1) - np.interp(T, T, np.arange(1, len(T) + 1))
    return T, S_T

T_vals, S_vals = compute_S_T(zeta_zeros)

# 🔹 **3. Visualisierung von S(T)**
plt.figure(figsize=(10,5))
plt.plot(T_vals, S_vals, color='blue', linewidth=0.7)
plt.xlabel("T")
plt.ylabel("S(T)")
plt.title("Numerische Berechnung von S(T)")
plt.grid(True)
plt.show()

# 🔹 **4. Fourier-Analyse von S(T)**
fft_vals = fft(S_vals)
freqs = np.fft.fftfreq(len(S_vals), d=(T_vals[1] - T_vals[0]))

plt.figure(figsize=(10,5))
plt.plot(freqs[:len(freqs)//2], np.abs(fft_vals[:len(fft_vals)//2]), color="green")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Stärke")
plt.title("Fourier-Analyse von S(T)")
plt.show()

# 🔹 **5. Wavelet-Transformation für lokale Frequenzanalyse**
scales = np.arange(1, 256)
coeffs, freqs = pywt.cwt(S_vals, scales, 'cmor')

plt.figure(figsize=(10,5))
plt.imshow(np.abs(coeffs), aspect='auto', extent=[T_vals[0], T_vals[-1], scales[-1], scales[0]], cmap="coolwarm")
plt.colorbar(label="Koeffizienten")
plt.xlabel("T")
plt.ylabel("Skalen (Frequenz)")
plt.title("Wavelet-Analyse von S(T)")
plt.show()

# 🔹 **6. Vergleich mit der Riemannschen Näherung**
def riemann_approximation(T):
    return (1 / np.pi) * np.angle(np.exp(1j * np.pi * np.log(T) / np.log(2)))

S_approx = riemann_approximation(T_vals)

plt.figure(figsize=(10,5))
plt.plot(T_vals, S_vals, label="Numerisches S(T)", color="blue", linewidth=0.7)
plt.plot(T_vals, S_approx, label="Riemann Approx.", linestyle="dashed", color="red", linewidth=0.7)
plt.xlabel("T")
plt.ylabel("S(T)")
plt.title("Vergleich von S(T) mit der Riemannschen Näherung")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 📌 Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import pywt  # 📌 Wavelet-Analyse
from scipy.fftpack import fft
from scipy.integrate import cumtrapz

# 📌 Nullstellen laden
filename = "/content/drive/MyDrive/zeros6.txt"
nullstellen = np.loadtxt(filename)

# 📌 Berechnung von S(T)
T_values = np.arange(1, len(nullstellen) + 1)
L_N = np.diff(nullstellen)  # Nullstellenabstände
S_T = np.insert(cumtrapz(L_N, T_values[:-1], initial=0), 0, 0)  # Integrierte Nullstellen-Abstände

# 📌 Trendentfernung mit polynomiale Anpassung
poly_degree = 3  # Niedrige Ordnung für schnelle Berechnung
p_coeff = np.polyfit(T_values, S_T, poly_degree)
trend = np.polyval(p_coeff, T_values)
residuen = S_T - trend  # Entfernte Trendfunktion

# 📌 Fourier-Analyse der Residuen
fft_values = np.abs(fft(residuen))
freqs = np.fft.fftfreq(len(residuen), d=(T_values[1] - T_values[0]))

# 📌 Wavelet-Analyse mit Daubechies-Wavelets (db4)
wavelet = "db4"
coeffs = pywt.wavedec(residuen, wavelet, level=6)  # 6 Ebenen für Detailauflösung
cA6, cD6, cD5, cD4, cD3, cD2, cD1 = coeffs  # Approximations- und Detailkoeffizienten

# 📌 📊 Ergebnisse plotten
plt.figure(figsize=(10,5))
plt.plot(T_values, S_T, label="S(T) Original", color="blue")
plt.plot(T_values, trend, label="Modellierter Trend", linestyle="dashed", color="red")
plt.xlabel("T")
plt.ylabel("S(T)")
plt.title("Numerische Berechnung von S(T) mit Trend")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(T_values, residuen, label="Trendentferntes S(T)", color="purple")
plt.xlabel("T")
plt.ylabel("Residuum von S(T)")
plt.title("Residuale Funktion nach Trendentfernung")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(freqs[:len(freqs)//2], fft_values[:len(fft_values)//2], color="green")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Stärke")
plt.title("Fourier-Analyse der detrendeten S(T) Funktion")
plt.grid()
plt.show()

# 📌 Wavelet-Koeffizienten plotten (Detailanalyse)
plt.figure(figsize=(10,5))
plt.plot(cD1, label="Detail Level 1", alpha=0.7)
plt.plot(cD2, label="Detail Level 2", alpha=0.7)
plt.plot(cD3, label="Detail Level 3", alpha=0.7)
plt.plot(cD4, label="Detail Level 4", alpha=0.7)
plt.plot(cD5, label="Detail Level 5", alpha=0.7)
plt.plot(cD6, label="Detail Level 6", alpha=0.7)
plt.xlabel("Koeffizienten-Index")
plt.ylabel("Wavelet-Koeffizienten")
plt.title("Wavelet-Analyse der detrendeten S(T) Funktion")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Parameter der Fibonacci-Freese-Formel
A = 0.0001
beta = 0.486
B = -1.6
w = 0.002
phi = -9005
C = -0.08
D = 7.8

# Definition der Differentialgleichung F'(N)
def dF_dN(N, F):
    return beta * F - B * w * N * np.cos(w * np.log(N + 1) + phi) + beta * B * np.sin(w * np.log(N + 1) + phi) \
           + C * (1 - beta * np.log(N + 1)) - D * (1 + beta / (N + 1))

# Lösungsbereich für N
N_min, N_max = 10, 10**7
N_values = np.logspace(1, 7, num=1000)  # Logarithmische Skalierung von 10 bis 10^7

# Anfangsbedingung für F(N) bei N=10
F0 = A * 10**beta + B * np.sin(w * np.log(10) + phi) + C * np.log(10) + D / 10

# Numerische Lösung mit Runge-Kutta (solve_ivp)
solution = solve_ivp(dF_dN, [N_min, N_max], [F0], t_eval=N_values, method='RK45')

# Plot der Lösung
plt.figure(figsize=(10, 5))
plt.plot(solution.t, solution.y[0], label="Numerische Lösung F(N)", color="blue")
plt.xlabel("N")
plt.ylabel("F(N)")
plt.xscale("log")
plt.title("Numerische Lösung der Fibonacci-Freese Differentialgleichung")
plt.legend()
plt.grid()
plt.show()
```
