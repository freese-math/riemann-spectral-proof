```python
# --- Bibliotheken ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# --- Dateipfade ---
beta_path = "/content/drive/MyDrive/beta_skala_kombi_repaired.csv"
zeros_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
primes_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv"

# --- Daten einlesen ---
beta_raw = pd.read_csv(beta_path).iloc[:, -1].astype(float).values
zeros = pd.read_csv(zeros_path, header=None).iloc[:, 0].astype(float).values
primes = pd.read_csv(primes_path, header=None).iloc[:, 0].astype(float).values

# --- Normierung ---
N = min(len(beta_raw), len(zeros), len(primes))
beta = beta_raw[:N]
beta_norm = beta / np.sum(beta)

gamma = zeros[:N]
logp = np.log(primes[:N])

# --- Vergleich: beta vs. Nullstellen ---
plt.figure(figsize=(8, 4))
plt.plot(beta_norm, label=r'$\beta(n)$', alpha=0.7)
plt.plot(gamma / np.max(gamma), label=r'$\gamma_n$ (skaliert)', alpha=0.7)
plt.title(r'Vergleich: $\beta(n)$ vs. Zeta-Nullstellen $\gamma_n$')
plt.xlabel("n")
plt.ylabel("Wert (skaliert)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Vergleich: beta vs. log(p) ---
plt.figure(figsize=(8, 4))
plt.plot(beta_norm, label=r'$\beta(n)$', alpha=0.7)
plt.plot(logp / np.max(logp), label=r'$\log(p_n)$ (skaliert)', alpha=0.7)
plt.title(r'Vergleich: $\beta(n)$ vs. $\log(p_n)$')
plt.xlabel("n")
plt.ylabel("Wert (skaliert)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Korrelation berechnen ---
corr_gamma = pearsonr(beta_norm, gamma)[0]
corr_logp = pearsonr(beta_norm, logp)[0]

print(f"Korrelation β vs. γ (Zeta): {corr_gamma:.6f}")
print(f"Korrelation β vs. log(p):   {corr_logp:.6f}")

# --- 3D Plot: beta vs. log(p) vs. gamma ---
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(beta_norm, logp / np.max(logp), gamma / np.max(gamma), color='darkblue')
ax.set_xlabel(r'$\beta(n)$')
ax.set_ylabel(r'$\log(p_n)$ (skaliert)')
ax.set_zlabel(r'$\gamma_n$ (skaliert)')
ax.set_title("3D-Struktur: β vs log(p) vs γ")
plt.tight_layout()
plt.show()

# --- PCA zur Dimensionsreduktion ---
X = np.vstack([beta_norm, logp, gamma]).T
X_std = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_std)

# --- PCA-Plot ---
plt.figure(figsize=(6, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], s=2, alpha=0.5)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA der Struktur [β, log(p), γ]")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Notwendige Bibliotheken ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from numpy.fft import fft, fftfreq

# --- Parameter ---
N = 10000  # Anzahl der Punkte
n = np.arange(1, N + 1)

# --- Analytisch abgeleitete Frequenzen (z. B. log(p), Hecke, Theta) ---
dominante_fk = np.array([
    9.994743e-06, 9.495006e-06, 8.995268e-06, 8.495531e-06,
    7.995794e-06, 7.496057e-06, 6.996320e-06, 6.496583e-06,
    5.996846e-06, 5.497109e-06, 4.997371e-06, 4.497634e-06,
    3.997897e-06, 3.498160e-06, 2.998423e-06, 2.498686e-06,
    1.998949e-06, 9.994743e-07, 1.499211e-06, 4.997371e-07
])
amplituden = np.array([
    1.82e8, 1.91e8, 2.00e8, 2.11e8, 2.23e8, 2.36e8, 2.51e8,
    2.68e8, 2.87e8, 3.09e8, 3.35e8, 3.66e8, 4.03e8, 4.47e8,
    5.02e8, 5.70e8, 6.54e8, 7.44e8, 7.45e8, 1.30e9
])
phasen = np.zeros_like(amplituden)

# --- Beta(n) rekonstruieren ---
beta_rekonstruiert = np.sum([
    A * np.sin(2 * np.pi * f * n + phi)
    for A, f, phi in zip(amplituden, dominante_fk, phasen)
], axis=0)

# --- FFT-Spektrum ---
fft_spectrum = np.abs(fft(beta_rekonstruiert))
frequencies = fftfreq(len(beta_rekonstruiert))

# --- Plot: FFT ---
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:N // 2], fft_spectrum[:N // 2])
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der rekonstruierten Beta-Skala")
plt.grid()
plt.tight_layout()
plt.show()

# --- Vergleich mit Original-Skala ---
# Optional: Originaldaten z.B. aus CSV laden
# beta_original = pd.read_csv("beta_skala_kombi_repaired.csv").iloc[:N, -1].values
# corr, _ = pearsonr(beta_rekonstruiert, beta_original)
# print(f"Korrelation mit Original-Skala: {corr:.6f}")
```

```python
# 📌 Bibliotheken
import numpy as np
import matplotlib.pyplot as plt

# 📌 Rekonstruierte Beta(n) (wie zuvor definiert)
N = 100  # Länge der Reihe
n = np.arange(1, N + 1)

# Drift + Frequenz-Anteile (angepasst, siehe vorheriger Code)
A_drift = 1e9
p_drift = 1.5
drift = A_drift / (n ** p_drift)

frequencies = np.array([1e-5, 5e-6, 1e-6])
amplitudes = np.array([2e8, 3e8, 4e8])

beta = drift.copy()
for f, A in zip(frequencies, amplitudes):
    beta += A * np.sin(2 * np.pi * f * n)

# 📌 L-Funktion aus Beta(n)
def L_beta(s, beta_vals, n_vals):
    terms = beta_vals / n_vals**s
    return np.sum(terms)

# 📌 Wertebereich im komplexen s
real_range = np.linspace(0.5, 4, 300)
L_vals_real = [L_beta(s, beta, n) for s in real_range]

# 📌 Plot
plt.figure(figsize=(8, 4))
plt.plot(real_range, np.real(L_vals_real), label="Re(Lβ(s))")
plt.plot(real_range, np.imag(L_vals_real), label="Im(Lβ(s))", linestyle="--")
plt.xlabel("Re(s)")
plt.ylabel("Lβ(s)")
plt.title("L-Funktion aus Beta(n): Lβ(s) = Σ β(n)/n^s")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib

# 📌 Importiere Bibliotheken
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 📌 Parameter: Maximale Anzahl Zeta-Nullstellen laden
max_zeros_user = 1000  # frei einstellbar

# 📌 Zeta-Nullstellen aus Datei laden
file_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt"  # Pfad anpassen

try:
    with open(file_path, "r") as f:
        zeta_zeros_full = np.array([float(line.strip()) for line in f.readlines() if line.strip()])

    # Überprüfen, ob genug Nullstellen vorhanden sind
    max_zeros_actual = len(zeta_zeros_full)
    if max_zeros_user > max_zeros_actual:
        print(f"⚠️ Nur {max_zeros_actual} Nullstellen gefunden. Verwende maximal mögliche Anzahl.")
        max_zeros_user = max_zeros_actual

    zeta_zeros = zeta_zeros_full[:max_zeros_user]
    print(f"✅ {max_zeros_user} Zeta-Nullstellen geladen!")

except Exception as e:
    print(f"❌ Fehler beim Laden der Datei: {e}")
    zeta_zeros = np.linspace(0, max_zeros_user, max_zeros_user)  # Fallback

# 📌 Hamilton-Operator konstruieren (Tridiagonal)
N = len(zeta_zeros)
H = np.diag(zeta_zeros) + np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-1), k=-1)

# 📌 Eigenwerte & Eigenvektoren berechnen
eigenvalues, eigenvectors = la.eigh(H)

# 📌 Selbstadjungiertheit prüfen
is_self_adjoint = np.allclose(H, H.T.conj(), atol=1e-10)
print(f"✅ Selbstadjungiert: {is_self_adjoint}")

# 📌 Zeige erste 10 Eigenwerte
print("Erste 10 Eigenwerte:")
print(np.round(eigenvalues[:10], 6))

# 📌 Spektralanalyse (Fourier)
fft_spectrum = np.fft.fft(eigenvalues)
frequencies = np.fft.fftfreq(N)

# 📌 Dominante Frequenzen extrahieren
dominant_indices = np.argsort(np.abs(fft_spectrum))[-5:]
dominant_frequencies = frequencies[dominant_indices]
dominant_amplitudes = np.abs(fft_spectrum[dominant_indices])

print("🔥 Dominante Frequenzen:", np.round(dominant_frequencies, 6))
print("🔥 Korrespondierende Amplituden:", np.round(dominant_amplitudes, 4))

# 📌 Korrelation & Fehler zur Zeta-Skala
correlation = np.corrcoef(eigenvalues, zeta_zeros)[0, 1]
euler_freese_error = np.sum((eigenvalues - zeta_zeros) ** 2) / len(eigenvalues)

print(f"✅ Korrelation mit Zeta-Nullstellen: {correlation:.5f}")
print(f"✅ Fehler in Euler-Freese-Identität: {euler_freese_error:.5f}")

# 📌 Plot: Eigenwerte
plt.figure(figsize=(7, 5))
plt.plot(eigenvalues, "bo-", markersize=3, label="Eigenwerte")
plt.xlabel("Index")
plt.ylabel("Eigenwert")
plt.title("Hamilton-Spektrum")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 📌 Plot: Fourier-Spektrum
plt.figure(figsize=(7, 5))
plt.plot(frequencies[:N//2], np.abs(fft_spectrum)[:N//2], "r-")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Spektralanalyse des Spektrums")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Bibliotheken ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.stats import pearsonr
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import os
import json

# --- Konfiguration ---
AUTO_MODE = True
SAVE_OUTPUTS = True
EXPORT_PATH = "/content/drive/MyDrive/MOS_Auswertung/"
BETA_CSV = "/content/drive/MyDrive/beta_skala_kombi_repaired.csv"
ZEROS_CSV = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
PRIMES_CSV = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv"
MAX_N = 2000000

# --- Daten vorbereiten ---
def load_and_prepare():
    beta_raw = pd.read_csv(BETA_CSV).iloc[:, -1].astype(float).values
    zeros = pd.read_csv(ZEROS_CSV, header=None).iloc[:, 0].astype(float).values

    df_primes = pd.read_csv(PRIMES_CSV, header=None)
    if isinstance(df_primes.iloc[0, 0], str):
        df_primes = df_primes.iloc[1:]
    primes = df_primes.iloc[:, 0].astype(float).values

    N = min(MAX_N, len(beta_raw), len(zeros), len(primes))
    beta = beta_raw[:N]
    beta_norm = beta / np.sum(beta)
    gamma = zeros[:N]
    logp = np.log(primes[:N])

    return beta_norm, logp, gamma

# --- Residuen berechnen ---
def compute_residuals(beta, logp, gamma):
    return {
        "r_beta": gamma - np.cumsum(beta),
        "r_logp": gamma - logp,
        "r_diff": np.cumsum(beta) - logp
    }

# --- Analyse + Plot ---
def analyze_and_plot(res, gamma, logp, beta):
    os.makedirs(EXPORT_PATH, exist_ok=True)
    for name, r in res.items():
        plt.figure(figsize=(8, 4))
        plt.plot(r, linewidth=0.5)
        plt.axhline(0, color='black', linestyle='--')
        plt.title(f"Residual: {name}")
        plt.grid(True)
        plt.tight_layout()
        if SAVE_OUTPUTS:
            plt.savefig(f"{EXPORT_PATH}{name}.png")
        plt.show()

        mean = np.mean(r)
        std = np.std(r)
        print(f"{name}: Mittelwert = {mean:.4f}, Std = {std:.4f}")

        # FFT
        fft_vals = np.abs(rfft(r - np.mean(r)))
        fft_freqs = rfftfreq(len(r), d=1)
        plt.figure(figsize=(8, 4))
        plt.plot(fft_freqs[:10000], fft_vals[:10000], linewidth=0.7)
        plt.title(f"FFT von {name}")
        plt.xlabel("Frequenz")
        plt.grid(True)
        plt.tight_layout()
        if SAVE_OUTPUTS:
            plt.savefig(f"{EXPORT_PATH}{name}_fft.png")
        plt.show()

# --- Statistik + PCA speichern ---
def export_results(beta, logp, gamma, res):
    corr1 = pearsonr(beta, gamma)[0]
    corr2 = pearsonr(beta, logp)[0]
    stats = {
        "correlation_beta_gamma": float(corr1),
        "correlation_beta_logp": float(corr2),
        "length": len(beta),
        "mean_r_beta": float(np.mean(res["r_beta"])),
        "mean_r_logp": float(np.mean(res["r_logp"])),
    }
    if SAVE_OUTPUTS:
        with open(f"{EXPORT_PATH}stats.json", "w") as f:
            json.dump(stats, f, indent=4)

    # PCA
    X = np.vstack([beta, logp, gamma]).T
    X_std = StandardScaler().fit_transform(X)
    X_pca = PCA(n_components=2).fit_transform(X_std)
    plt.figure(figsize=(6, 5))
    plt.scatter(X_pca[:, 0], X_pca[:, 1], s=2, alpha=0.5)
    plt.title("PCA [β, log(p), γ]")
    plt.grid(True)
    plt.tight_layout()
    if SAVE_OUTPUTS:
        plt.savefig(f"{EXPORT_PATH}pca_plot.png")
    plt.show()

# --- Main ---
if AUTO_MODE:
    beta_norm, logp, gamma = load_and_prepare()
    residuals = compute_residuals(beta_norm, logp, gamma)
    analyze_and_plot(residuals, gamma, logp, beta_norm)
    export_results(beta_norm, logp, gamma, residuals)
```
