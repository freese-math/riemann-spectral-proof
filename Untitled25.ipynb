```python
# 📌 Schritt 1: Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

# 📌 Schritt 2: Parameter & Diskretisierung definieren
N = 1000  # Anzahl der Gitterpunkte
L = 10  # Länge des Intervalls
x = np.linspace(-L, L, N)
dx = x[1] - x[0]

# 📌 Schritt 3: Pöschl-Teller-Potential definieren
V0 = 10  # Tiefe des Potentials
alpha = 1.0  # Breite des Potentials
V_ptp = -V0 / np.cosh(alpha * x) ** 2  # Effektives Potential

# 📌 Schritt 4: Diskretisierung des Schrödinger-Operators (Hamilton-Matrix)
kinetic = -0.5 / dx**2 * np.ones(N-1)  # Kinetischer Term (Laplace-Operator)
potential = V_ptp + 0.5 / dx**2  # Gesamtpotential
H_diag = potential  # Diagonalelemente
H_offdiag = kinetic  # Nebendiagonalelemente

# 📌 Schritt 5: Numerische Eigenwerte berechnen
eigenvalues_ptp, eigenvectors = eigh_tridiagonal(H_diag, H_offdiag)

# 📌 Schritt 6: Vergleich mit bekannten Riemann-Zeta-Nullstellen
zeta_zeros = np.array([14.134725, 21.02204, 25.01086, 30.42487, 32.93506, 37.58618, 40.91872, 43.32707])

# 📌 Schritt 7: Visualisierung des Vergleichs
plt.figure(figsize=(8, 5))
plt.plot(eigenvalues_ptp[:len(zeta_zeros)], zeta_zeros, 'bo', label="Zeta-Nullstellen vs. PTP")
plt.plot([0, max(eigenvalues_ptp[:len(zeta_zeros)])], [0, max(zeta_zeros)], 'r--', label="Ideal")
plt.xlabel("Berechnete Eigenwerte (Pöschl-Teller)")
plt.ylabel("Zeta-Nullstellen")
plt.legend()
plt.title("Vergleich: Schrödinger-Operator vs. Riemannsche Zeta-Nullstellen")
plt.grid()
plt.show()
```

```python
# 🚀 Colab-ready Code zur Analyse der Fibonacci-Freese-Formel (FFF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import zeta
from scipy.fft import fft, fftfreq
from scipy.linalg import eigvalsh
import mpmath as mp

# 🎯 1. Berechnung der ersten N Nullstellen der Riemannschen Zeta-Funktion
def compute_zeta_zeros(N):
    mp.dps = 50  # Hohe Genauigkeit für Nullstellen
    zeros = [mp.zetazero(n).imag for n in range(1, N+1)]
    return np.array(zeros)

# 🎯 2. Berechnung der Eigenwerte eines Schrödinger-Operators mit Pöschl-Teller-Potential
def poeschl_teller_eigenvalues(N, alpha=2.010362):
    """Berechnet die ersten N Eigenwerte für ein Pöschl-Teller-Potential"""
    return -np.array([(alpha - n) ** 2 for n in range(1, N+1)])

# 🎯 3. Vergleich der Zeta-Nullstellen mit den Pöschl-Teller-Eigenwerten
N = 50  # Anzahl der zu berechnenden Nullstellen/Eigenwerte
zeta_zeros = compute_zeta_zeros(N)
poeschl_eigenvalues = poeschl_teller_eigenvalues(N)

# 📈 4. Visualisierung: Vergleich Schrödinger-Eigenwerte vs. Zeta-Nullstellen
plt.figure(figsize=(8,5))
plt.scatter(poeschl_eigenvalues, zeta_zeros, color='blue', label="Zeta-Nullstellen vs. PTP")
plt.plot(poeschl_eigenvalues, -poeschl_eigenvalues / 2, 'r--', label="Ideale Skalierung")
plt.xlabel("Berechnete Eigenwerte (Pöschl-Teller)")
plt.ylabel("Zeta-Nullstellen")
plt.title("Vergleich: Schrödinger-Operator vs. Riemannsche Zeta-Nullstellen")
plt.legend()
plt.show()

# 🎯 5. Fourier-Analyse der Nullstellen-Abstände
delta_zeta = np.diff(zeta_zeros)  # Abstände zwischen den Nullstellen
frequencies = fftfreq(N-1, d=1)[:N//2]
spectrum = np.abs(fft(delta_zeta))[:N//2]

# 📈 6. Visualisierung: Fourier-Spektrum der Nullstellen-Abstände
plt.figure(figsize=(10,5))
plt.plot(frequencies, spectrum, 'b-', label="Fourier-Spektrum")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstände")
plt.legend()
plt.show()

# 🎯 7. Test der Fibonacci-Freese-Formel (FFF)
def fibonacci_freese(N, beta=0.912576):
    """Berechnet L(N) nach der FFF: L(N) ≈ N^β"""
    return N ** beta

# Berechnung der Kohärenzlänge L(N) nach FFF
L_FFF = fibonacci_freese(np.arange(1, N+1))

# 📈 8. Visualisierung: FFF-Fit mit tatsächlichen Nullstellenabständen
plt.figure(figsize=(10,5))
plt.plot(np.arange(1, N+1), L_FFF, 'r--', label=f"FFF Fit: α={2.010362}, β={0.912576}")
plt.plot(np.arange(1, N), np.cumsum(delta_zeta), 'b-', label="Kohärenzlänge L(N)")
plt.xlabel("N")
plt.ylabel("L(N)")
plt.title("Fibonacci-Freese-Formel Fit")
plt.legend()
plt.yscale('log')  # Log-Skalierung für besseren Vergleich
plt.xscale('log')
plt.show()

# 🎯 9. Monte-Carlo-Simulation zur Stabilität der Fibonacci-Freese-Formel
num_samples = 1000
monte_carlo_results = []
for _ in range(num_samples):
    noisy_data = delta_zeta + np.random.normal(0, 0.005, size=len(delta_zeta))
    fft_noise = np.abs(fft(noisy_data))[:N//2]
    monte_carlo_results.append(np.max(fft_noise))

# 📈 10. Visualisierung: Monte-Carlo-Stabilitätstest der FFF
plt.figure(figsize=(8,5))
plt.hist(monte_carlo_results, bins=50, color='purple', alpha=0.75)
plt.xlabel("Fehler |Zeta - FFF|")
plt.ylabel("Häufigkeit")
plt.title("Monte-Carlo-Stabilitätstest der FFF (CPU)")
plt.show()
```

```python
# Installiere ggf. notwendige Bibliotheken
!pip install numpy matplotlib scipy

# 📌 Importieren der benötigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 📌 Fibonacci-Freese-Formel (FFF)
def fff_formula(N, alpha, beta, gamma):
    return alpha * N**beta + gamma

# 📌 Monte-Carlo-Simulation zur Fehlerbewertung
def monte_carlo_simulation(N_samples=1000):
    np.random.seed(42)  # Reproduzierbarkeit
    N_values = np.logspace(0.5, 6, N_samples)  # Testwerte für N

    # 📌 Simulierte "echte" Kohärenzlängen (mit zufälligem Rauschen)
    true_L_N = 2.010362 * N_values**0.912576 + np.random.normal(0, 5, len(N_values))

    # 📌 Fitting der FFF
    popt, pcov = curve_fit(fff_formula, N_values, true_L_N, p0=[2, 0.9, 0])

    # 📌 Fehlerberechnung
    predicted_L_N = fff_formula(N_values, *popt)
    errors = np.abs(true_L_N - predicted_L_N)

    avg_error = np.mean(errors)
    std_dev_error = np.std(errors)
    max_error = np.max(errors)
    min_error = np.min(errors)

    # 📌 Ergebnisse anzeigen
    print(f"📊 Fehlerstatistik:")
    print(f"🔹 Durchschnittlicher Fehler: {avg_error:.6f}")
    print(f"🔹 Standardabweichung des Fehlers: {std_dev_error:.6f}")
    print(f"🔹 Maximaler Fehler: {max_error:.6f}")
    print(f"🔹 Minimaler Fehler: {min_error:.6f}")

    # 📌 Histogramm des Fehlers plotten
    plt.figure(figsize=(8,5))
    plt.hist(errors, bins=50, color="purple", alpha=0.7)
    plt.xlabel("Fehler |Zeta - FFF|")
    plt.ylabel("Häufigkeit")
    plt.title("Monte-Carlo-Stabilitätstest der FFF")
    plt.show()

    # 📌 Angepasste Parameter für FFF ausgeben
    print("\n📌 Angepasste Parameter für FFF:")
    print(f"🔹 α = {popt[0]:.6f}")
    print(f"🔹 β = {popt[1]:.6f}")
    print(f"🔹 γ (Offset) = {popt[2]:.6f}")

# 📌 Simulation ausführen
monte_carlo_simulation()
```

```python
# 📌 Installiere ggf. benötigte Bibliotheken
!pip install numpy matplotlib scipy sympy mpmath requests

# 📌 Importiere benötigte Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import requests
from scipy.fftpack import fft
from sympy import prime
from mpmath import zetazero

# 📌 **Parameter & Einstellungen**
N = 2000000  # Anzahl der Nullstellen (2 Millionen)
alpha_FFF = 2.010512
beta_FFF = 0.91257
gamma_FFF = -0.134437

USE_RAINBOW_TABLE = False  # 🌈 Rainbow Table optional verwenden

# 📌 **📊 Funktionen zum Laden der Daten**
def load_lmfdb_zeros():
    """ Lädt die ersten 100.000 Nullstellen aus LMFDB (Web-Quelle). """
    url = "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"
    response = requests.get(url)
    zeros = [float(line.split()[-1]) for line in response.text.split("\n") if line and line[0].isdigit()]
    return np.array(zeros)

def load_odlyzko_zeros(file_path):
    """ Lädt Nullstellen aus der Odlyzko-Datei. """
    with open(file_path, "r") as f:
        zeros = [float(line.strip()) for line in f if line.strip()]
    return np.array(zeros)

def load_primes(file_path):
    """ Lädt Primzahlen aus Datei. """
    with open(file_path, "r") as f:
        primes = [int(line.strip()) for line in f if line.strip()]
    return np.array(primes)

# 📌 **📊 Lade die Daten**
lmfdb_zeros = load_lmfdb_zeros()
odlyzko_zeros = load_odlyzko_zeros("/content/drive/MyDrive/zeros6.txt")
primes = load_primes("/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen1mio.txt")

if USE_RAINBOW_TABLE:
    rainbow_zeros = load_odlyzko_zeros("/content/drive/MyDrive/rainbow_table_10M.txt")

# 📌 **1️⃣ Berechnung der Nullstellen-Abstände**
def compute_gaps(data):
    """ Berechnet Abstände zwischen aufeinanderfolgenden Nullstellen. """
    return np.diff(data)

zeta_gaps = compute_gaps(odlyzko_zeros)
prime_gaps = compute_gaps(primes[:len(zeta_gaps)])  # Nur gleich viele Werte nehmen

# 📌 **2️⃣ Fourier-Analyse der Nullstellen-Abstände**
def compute_fft(data):
    """ Berechnet die Fourier-Transformation der Abstände. """
    spectrum = np.abs(fft(data))[:len(data)//2]
    frequencies = np.linspace(0, 0.5, len(spectrum))
    return frequencies, spectrum

freqs, zeta_spectrum = compute_fft(zeta_gaps)

# 📌 **3️⃣ Fibonacci-Sequenz als Skalen-Korrektur**
def fibonacci_sequence(n):
    """ Generiert die ersten n Fibonacci-Zahlen. """
    fib = [1, 1]
    for _ in range(n-2):
        fib.append(fib[-1] + fib[-2])
    return np.array(fib)

fib_seq = fibonacci_sequence(100)

# 📌 **4️⃣ Kohärenzlänge & Fibonacci-Freese-Formel**
def fff_kohärenzlänge(N, alpha, beta, gamma):
    """ Berechnet die FFF-Kohärenzlänge mit Theta-Siegel-Korrektur. """
    return alpha * N ** beta + gamma

N_values = np.arange(1, N+1)
L_N = fff_kohärenzlänge(N_values, alpha_FFF, beta_FFF, gamma_FFF)

# 📌 **📊 5️⃣ Plots erstellen**

# 🔹 Fibonacci-Sequenz als Skalen-Korrektur
plt.figure(figsize=(10,5))
plt.plot(fib_seq, 'r--', label="Fibonacci-Sequenz")
plt.xlabel("Index N")
plt.ylabel("Fibonacci-Wert")
plt.title("Fibonacci-Sequenz als Skalen-Korrektur")
plt.legend()
plt.grid()
plt.show()

# 🔹 Fibonacci-Freese-Formel mit Theta-Korrektur
plt.figure(figsize=(10,5))
plt.loglog(N_values, L_N, 'r--', label=f"FFF Fit: α={alpha_FFF}, β={beta_FFF}, γ={gamma_FFF}")
plt.loglog(N_values, np.sqrt(N_values), 'k-', label="Theta-Siegel-Funktion")
plt.xlabel("N")
plt.ylabel("Kohärenzlänge L(N)")
plt.title("Fibonacci-Freese-Formel Fit mit Theta-Korrektur")
plt.legend()
plt.grid()
plt.show()

# 🔹 Fourier-Analyse der Nullstellen-Abstände
plt.figure(figsize=(10,5))
plt.plot(freqs, zeta_spectrum, 'b', label="Fourier-Spektrum")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstände (echte Daten)")
plt.legend()
plt.grid()
plt.show()

# 📌 **📊 6️⃣ Monte-Carlo-Fehleranalyse**
def monte_carlo_error_analysis(true_values, predicted_values):
    """ Berechnet Fehlerstatistik für Monte-Carlo-Analyse. """
    errors = np.abs(true_values - predicted_values)
    return {
        "Durchschnittlicher Fehler": np.mean(errors),
        "Standardabweichung": np.std(errors),
        "Maximaler Fehler": np.max(errors),
        "Minimaler Fehler": np.min(errors),
    }

error_stats = monte_carlo_error_analysis(zeta_gaps[:len(L_N)], L_N[:len(zeta_gaps)])

# 📊 Fehlerstatistik ausgeben
print("\n📊 Fehlerstatistik:")
for key, value in error_stats.items():
    print(f"🔹 {key}: {value:.6f}")
```

```python
import numpy as np

# 📌 Pfade der Dateien
file_generated = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen.txt"
file_existing = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen1mio.txt"

# 📌 1️⃣ **Datei mit Komma-getrennten Zahlen bereinigen**
def clean_comma_separated_file(input_path, output_path):
    with open(input_path, "r") as file:
        content = file.read()

    # Ersetze Kommas durch Zeilenumbrüche und speichere als neue Datei
    cleaned_numbers = content.replace(",", "\n")

    with open(output_path, "w") as file:
        file.write(cleaned_numbers)

# 📌 2️⃣ **Vergleich der beiden Primzahl-Dateien**
def compare_prime_files(file1, file2):
    primes1 = np.loadtxt(file1, dtype=int)
    primes2 = np.loadtxt(file2, dtype=int)

    if np.array_equal(primes1, primes2):
        print("✅ Die beiden Primzahl-Listen sind identisch!")
    else:
        print("⚠️ Die Primzahlen unterscheiden sich!")
        diff = set(primes1) ^ set(primes2)
        print(f"Unterschiede (max. 10 Einträge): {list(diff)[:10]}")

# 📌 Bereinigung ausführen
cleaned_file = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen_cleaned.txt"
clean_comma_separated_file(file_generated, cleaned_file)

# 📌 Dateien vergleichen
compare_prime_files(cleaned_file, file_existing)
```

```python
# 📌 **1️⃣ Bibliotheken importieren**
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import pandas as pd

# 📌 **2️⃣ Datenpfade setzen**
primes_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen.txt"
zeros_path = "/content/drive/MyDrive/zeros6.txt"

# 📌 **3️⃣ Daten einlesen**
def load_primes(file_path):
    with open(file_path, "r") as f:
        primes = np.array([int(line.strip()) for line in f])
    return primes

def load_zeros(file_path):
    with open(file_path, "r") as f:
        zeros = np.array([float(line.strip()) for line in f])
    return zeros

primes = load_primes(primes_path)
zeros = load_zeros(zeros_path)

# **Daten validieren**
assert len(primes) == 1_000_000, f"⚠️ Unerwartete Anzahl an Primzahlen: {len(primes)}"
assert len(zeros) == 2_001_052, f"⚠️ Unerwartete Anzahl an Nullstellen: {len(zeros)}"

print("✅ 1,000,000 Primzahlen geladen.")
print("✅ 2,001,052 Nullstellen geladen.")

# 📌 **4️⃣ Abstandsberechnungen**
abstaende = np.diff(zeros)

# 📌 **5️⃣ Fibonacci-Freese-Formel (FFF)**
def FFF(N, alpha, beta, gamma):
    return alpha * N**beta + gamma

# 📌 **6️⃣ Curve-Fitting zur Bestimmung der FFF-Parameter**
N_values = np.arange(1, len(zeros) + 1)
y_values = np.cumsum(abstaende)  # Kumulative Abstände

popt_fff, _ = opt.curve_fit(FFF, N_values, y_values, p0=[1.88, 0.916, 2488])
alpha_opt, beta_opt, gamma_opt = popt_fff

# 📌 **7️⃣ Alternative Approximationen**
def log_poly(x, a, b, c):
    return a + b * np.log(x) + c * (np.log(x))**2

def exp_fit(x, a, b, c):
    return a * np.exp(b * x) + c

popt_log, _ = opt.curve_fit(log_poly, N_values, np.log(y_values))
popt_exp, _ = opt.curve_fit(exp_fit, N_values, y_values, maxfev=10000)

# 📌 **8️⃣ Fehlerberechnung**
y_pred_fff = FFF(N_values, *popt_fff)
y_pred_log = np.exp(log_poly(N_values, *popt_log))
y_pred_exp = exp_fit(N_values, *popt_exp)

errors_fff = y_values - y_pred_fff
errors_log = y_values - y_pred_log
errors_exp = y_values - y_pred_exp

error_stats = {
    "Modell": ["FFF", "Log-Poly", "Exponentiell"],
    "Durchschnittlicher Fehler": [np.mean(errors_fff), np.mean(errors_log), np.mean(errors_exp)],
    "Standardabweichung": [np.std(errors_fff), np.std(errors_log), np.std(errors_exp)],
    "Maximaler Fehler": [np.max(errors_fff), np.max(errors_log), np.max(errors_exp)]
}

df_errors = pd.DataFrame(error_stats)

# 📌 **9️⃣ Visualisierungen**
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Echte Werte", color="black")
plt.plot(N_values, y_pred_fff, "--", label="FFF Fit", color="red")
plt.plot(N_values, y_pred_log, "--", label="Log-Poly Fit", color="blue")
plt.plot(N_values, y_pred_exp, "--", label="Exponentieller Fit", color="green")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("L(N)")
plt.legend()
plt.title("Vergleich alternativer Approximationen")
plt.show()

# 📊 Fehlerverteilung
plt.figure(figsize=(10,5))
plt.hist(errors_fff, bins=100, color="red", alpha=0.6, label="FFF Fehler")
plt.hist(errors_log, bins=100, color="blue", alpha=0.6, label="Log-Poly Fehler")
plt.hist(errors_exp, bins=100, color="green", alpha=0.6, label="Exponentieller Fehler")
plt.xlabel("Fehler")
plt.ylabel("Häufigkeit")
plt.legend()
plt.title("Fehlerverteilung der Approximationen")
plt.show()

# 📌 **10️⃣ Pearson-Korrelation zwischen Primzahlen & Nullstellen**
from scipy.stats import pearsonr
corr, _ = pearsonr(primes[:len(zeros)], zeros)

print(f"📌 Pearson-Korrelation (Primzahlen vs. Nullstellen): {corr:.6f}")

# 📌 **Fehlerstatistik ausgeben**
import ace_tools as tools
tools.display_dataframe_to_user(name="Fehleranalyse alternative Approximationen", dataframe=df_errors)
```
