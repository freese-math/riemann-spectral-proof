```python
#  Schritt 1: Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

#  Schritt 2: Parameter & Diskretisierung definieren
N = 1000  # Anzahl der Gitterpunkte
L = 10  # Lnge des Intervalls
x = np.linspace(-L, L, N)
dx = x[1] - x[0]

#  Schritt 3: Pschl-Teller-Potential definieren
V0 = 10  # Tiefe des Potentials
alpha = 1.0  # Breite des Potentials
V_ptp = -V0 / np.cosh(alpha * x) ** 2  # Effektives Potential

#  Schritt 4: Diskretisierung des Schrdinger-Operators (Hamilton-Matrix)
kinetic = -0.5 / dx**2 * np.ones(N-1)  # Kinetischer Term (Laplace-Operator)
potential = V_ptp + 0.5 / dx**2  # Gesamtpotential
H_diag = potential  # Diagonalelemente
H_offdiag = kinetic  # Nebendiagonalelemente

#  Schritt 5: Numerische Eigenwerte berechnen
eigenvalues_ptp, eigenvectors = eigh_tridiagonal(H_diag, H_offdiag)

#  Schritt 6: Vergleich mit bekannten Riemann-Zeta-Nullstellen
zeta_zeros = np.array([14.134725, 21.02204, 25.01086, 30.42487, 32.93506, 37.58618, 40.91872, 43.32707])

#  Schritt 7: Visualisierung des Vergleichs
plt.figure(figsize=(8, 5))
plt.plot(eigenvalues_ptp[:len(zeta_zeros)], zeta_zeros, 'bo', label="Zeta-Nullstellen vs. PTP")
plt.plot([0, max(eigenvalues_ptp[:len(zeta_zeros)])], [0, max(zeta_zeros)], 'r--', label="Ideal")
plt.xlabel("Berechnete Eigenwerte (Pschl-Teller)")
plt.ylabel("Zeta-Nullstellen")
plt.legend()
plt.title("Vergleich: Schrdinger-Operator vs. Riemannsche Zeta-Nullstellen")
plt.grid()
plt.show()
```

```python
#  Colab-ready Code zur Analyse der Fibonacci-Freese-Formel (FFF)
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import zeta
from scipy.fft import fft, fftfreq
from scipy.linalg import eigvalsh
import mpmath as mp

#  1. Berechnung der ersten N Nullstellen der Riemannschen Zeta-Funktion
def compute_zeta_zeros(N):
    mp.dps = 50  # Hohe Genauigkeit fr Nullstellen
    zeros = [mp.zetazero(n).imag for n in range(1, N+1)]
    return np.array(zeros)

#  2. Berechnung der Eigenwerte eines Schrdinger-Operators mit Pschl-Teller-Potential
def poeschl_teller_eigenvalues(N, alpha=2.010362):
    """Berechnet die ersten N Eigenwerte fr ein Pschl-Teller-Potential"""
    return -np.array([(alpha - n) ** 2 for n in range(1, N+1)])

#  3. Vergleich der Zeta-Nullstellen mit den Pschl-Teller-Eigenwerten
N = 50  # Anzahl der zu berechnenden Nullstellen/Eigenwerte
zeta_zeros = compute_zeta_zeros(N)
poeschl_eigenvalues = poeschl_teller_eigenvalues(N)

#  4. Visualisierung: Vergleich Schrdinger-Eigenwerte vs. Zeta-Nullstellen
plt.figure(figsize=(8,5))
plt.scatter(poeschl_eigenvalues, zeta_zeros, color='blue', label="Zeta-Nullstellen vs. PTP")
plt.plot(poeschl_eigenvalues, -poeschl_eigenvalues / 2, 'r--', label="Ideale Skalierung")
plt.xlabel("Berechnete Eigenwerte (Pschl-Teller)")
plt.ylabel("Zeta-Nullstellen")
plt.title("Vergleich: Schrdinger-Operator vs. Riemannsche Zeta-Nullstellen")
plt.legend()
plt.show()

#  5. Fourier-Analyse der Nullstellen-Abstnde
delta_zeta = np.diff(zeta_zeros)  # Abstnde zwischen den Nullstellen
frequencies = fftfreq(N-1, d=1)[:N//2]
spectrum = np.abs(fft(delta_zeta))[:N//2]

#  6. Visualisierung: Fourier-Spektrum der Nullstellen-Abstnde
plt.figure(figsize=(10,5))
plt.plot(frequencies, spectrum, 'b-', label="Fourier-Spektrum")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstnde")
plt.legend()
plt.show()

#  7. Test der Fibonacci-Freese-Formel (FFF)
def fibonacci_freese(N, beta=0.912576):
    """Berechnet L(N) nach der FFF: L(N)  N^"""
    return N ** beta

# Berechnung der Kohrenzlnge L(N) nach FFF
L_FFF = fibonacci_freese(np.arange(1, N+1))

#  8. Visualisierung: FFF-Fit mit tatschlichen Nullstellenabstnden
plt.figure(figsize=(10,5))
plt.plot(np.arange(1, N+1), L_FFF, 'r--', label=f"FFF Fit: ={2.010362}, ={0.912576}")
plt.plot(np.arange(1, N), np.cumsum(delta_zeta), 'b-', label="Kohrenzlnge L(N)")
plt.xlabel("N")
plt.ylabel("L(N)")
plt.title("Fibonacci-Freese-Formel Fit")
plt.legend()
plt.yscale('log')  # Log-Skalierung fr besseren Vergleich
plt.xscale('log')
plt.show()

#  9. Monte-Carlo-Simulation zur Stabilitt der Fibonacci-Freese-Formel
num_samples = 1000
monte_carlo_results = []
for _ in range(num_samples):
    noisy_data = delta_zeta + np.random.normal(0, 0.005, size=len(delta_zeta))
    fft_noise = np.abs(fft(noisy_data))[:N//2]
    monte_carlo_results.append(np.max(fft_noise))

#  10. Visualisierung: Monte-Carlo-Stabilittstest der FFF
plt.figure(figsize=(8,5))
plt.hist(monte_carlo_results, bins=50, color='purple', alpha=0.75)
plt.xlabel("Fehler |Zeta - FFF|")
plt.ylabel("Hufigkeit")
plt.title("Monte-Carlo-Stabilittstest der FFF (CPU)")
plt.show()
```

```python
#  1 **Importe & Konfiguration**
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from scipy.stats import pearsonr
from scipy.fftpack import fft

#  2 **Datenpfade**
primes_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen.txt"
zeros_path = "/content/drive/MyDrive/zeros6.txt"

#  3 **Daten laden**
def load_primes(file_path):
    with open(file_path, "r") as f:
        primes = np.array([int(line.strip()) for line in f])
    return primes

def load_zeros(file_path):
    with open(file_path, "r") as f:
        zeros = np.array([float(line.strip()) for line in f])
    return zeros

#  4 **Laden & Validierung**
primes = load_primes(primes_path)
zeros = load_zeros(zeros_path)

assert len(primes) == 1_000_000, f" Primzahlenanzahl unerwartet: {len(primes)}"
assert len(zeros) == 2_001_052, f" Nullstellenanzahl unerwartet: {len(zeros)}"

print(" 1,000,000 Primzahlen geladen.")
print(" 2,001,052 Nullstellen geladen.")

#  5 **Abstandsberechnung**
primes_abstaende = np.diff(primes)
zeros_abstaende = np.diff(zeros)

#  6 **Fourier-Analyse der Nullstellen-Abstnde**
fft_spektrum = np.abs(fft(zeros_abstaende))[:len(zeros_abstaende)//2]
frequenzen = np.linspace(0, 0.5, len(fft_spektrum))

plt.figure(figsize=(10,5))
plt.plot(frequenzen, fft_spektrum, 'b', label="Fourier-Spektrum")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstnde")
plt.legend()
plt.grid()
plt.show()

#  7 **FFF-Modell & Curve-Fitting**
def FFF(N, alpha, beta, gamma):
    return alpha * N**beta + gamma

N_values = np.arange(1, len(zeros_abstaende) + 1)
popt, _ = opt.curve_fit(FFF, N_values, np.cumsum(zeros_abstaende), p0=[1.88, 0.916, 2488])

alpha_opt, beta_opt, gamma_opt = popt

plt.figure(figsize=(10,5))
plt.loglog(N_values, FFF(N_values, alpha_opt, beta_opt, gamma_opt), 'r--', label=f"FFF Fit: ={alpha_opt:.6f}, ={beta_opt:.6f}, ={gamma_opt:.6f}")
plt.xlabel("N")
plt.ylabel("L(N)")
plt.title("Fibonacci-Freese-Formel Fit")
plt.legend()
plt.grid()
plt.show()

#  8 **Fehleranalyse & Histogramm**
fehler = np.cumsum(zeros_abstaende) - FFF(N_values, alpha_opt, beta_opt, gamma_opt)

plt.figure(figsize=(10,5))
plt.hist(fehler, bins=50, color='blue', alpha=0.7, label="Fehlerverteilung")
plt.axvline(np.mean(fehler), color='r', linestyle='dashed', linewidth=2, label=" Fehler")
plt.axvline(np.max(fehler), color='k', linestyle='dotted', linewidth=2, label="Max Fehler")
plt.xlabel("Fehler")
plt.ylabel("Hufigkeit")
plt.title("Histogramm der Fehlerwerte")
plt.legend()
plt.grid()
plt.show()

#  9 **Residualanalyse**
plt.figure(figsize=(10,5))
plt.scatter(N_values, fehler, s=1, c='blue', label="Residuen")
plt.axhline(0, color='k', linestyle='dashed')
plt.xlabel("N")
plt.ylabel("Residuen (Fehler)")
plt.title("Residualanalyse der FFF (Fehlerverteilung)")
plt.legend()
plt.grid()
plt.show()

#   **Korrelation Primzahlen vs. Zeta-Nullstellen**
plt.figure(figsize=(10,5))
plt.scatter(primes[:len(zeros)], zeros[:len(primes)], c='red', s=1, label="Primzahlen vs. Zeta-Nullstellen")
plt.xlabel("Primzahlen")
plt.ylabel("Zeta-Nullstellen")
plt.title("Primzahlen vs. Zeta-Nullstellen Korrelation")
plt.legend()
plt.grid()
plt.show()

#  11 **Pearson-Korrelation**
corr, _ = pearsonr(primes[:len(zeros)], zeros[:len(primes)])
print(f" Pearson-Korrelation (Primzahlen vs. Nullstellen): {corr:.6f}")

#  **Ergebniszusammenfassung**
print("\n Optimierte Parameter fr FFF:")
print(f"  = {alpha_opt:.6f}")
print(f"  = {beta_opt:.6f}")
print(f"  = {gamma_opt:.6f}")

print("\n Fehlerstatistik:")
print(f" Durchschnittlicher Fehler: {np.mean(fehler):.6f}")
print(f" Standardabweichung: {np.std(fehler):.6f}")
print(f" Maximaler Fehler: {np.max(fehler):.6f}")

print(f"\n Pearson-Korrelation (Primzahlen vs. Nullstellen): {corr:.6f}")
```
