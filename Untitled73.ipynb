```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import mean_squared_error, mean_absolute_error

# ======================
# 1. Daten einlesen
# ======================

# Google Drive ist bereits gemountet
file_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
data = pd.read_csv(file_path, header=None)
gamma = data.iloc[:, 0].values  # Imaginrteile der Nullstellen
rho = 0.5 + 1j * gamma          # komplexe Nullstellen

# ======================
# 2. Empirische Beta(n): z.B. Frequenzabstand
# ======================

n = np.arange(1, len(gamma))             # Index n
beta_emp = np.diff(gamma)                # Nherung: (n)  Abstand zwischen _n+1 - _n
rho_emp = rho[:-1]                       # passende Lnge zu (n)
product = beta_emp * rho_emp             # komplexes Produkt (n) * (n)

# ======================
# 3. Fit: Betrag des Produkts ~  n^
# ======================

def model(n, alpha, delta):
    return alpha * n**delta

product_abs = np.abs(product)
params, _ = curve_fit(model, n[:len(product_abs)], product_abs, p0=[1.0, 1.0])

alpha_fit, delta_fit = params
print(f"Gefittet:  = {alpha_fit:.6f},  = {delta_fit:.6f}")

# ======================
# 4. Plot: Vergleich von Produkt und Fit
# ======================

plt.figure(figsize=(12, 6))
plt.plot(n[:len(product_abs)], product_abs, label=r'$|\beta(n)\cdot\rho_n|$ (empirisch)', lw=2)
plt.plot(n[:len(product_abs)], model(n[:len(product_abs)], *params), '--', label=r'Fit: $\alpha n^\delta$', lw=2)
plt.xlabel('n')
plt.ylabel(r'$|\beta(n)\cdot\rho_n|$')
plt.title('Produktstruktur: Vergleich mit Potenzansatz')
plt.legend()
plt.grid(True)
plt.show()

# ======================
# 5. Fehlermae
# ======================

mse = mean_squared_error(product_abs, model(n[:len(product_abs)], *params))
rmse = np.sqrt(mse)
mae = mean_absolute_error(product_abs, model(n[:len(product_abs)], *params))

print("Fehlermetriken:")
print(f"MSE  = {mse:.6f}")
print(f"RMSE = {rmse:.6f}")
print(f"MAE  = {mae:.6f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# ========================
# 1. Zeta-Nullstellen laden
# ========================
file_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
data = pd.read_csv(file_path, header=None)
gamma = data.iloc[:, 0].values

# ========================
# 2. Arbeitsbereich vorbereiten
# ========================
n = np.arange(2, 1002)
gamma_n = gamma[1:1001]

# ========================
# 3. Parametrisierbare Beta-Modellfunktion
# params = [b1, f1, phi1, b2, f2, phi2, b3, f3, phi3, c, d]
def beta_model(n, params):
    b1, f1, phi1 = params[0:3]
    b2, f2, phi2 = params[3:6]
    b3, f3, phi3 = params[6:9]
    c, d = params[9:11]

    A = 0.003652
    alpha = 1.336085

    return (
        A * n**(-alpha)
        + b1 * np.cos(2 * np.pi * f1 * n + phi1)
        + b2 * np.cos(2 * np.pi * f2 * n + phi2)
        + b3 * np.cos(2 * np.pi * f3 * n + phi3)
        + c * n**d
    )

# ========================
# 4. Zielfunktion: RMSE der Realteile
# ========================
def objective(params):
    beta_vals = beta_model(n, params)
    phase = np.exp(1j * beta_vals)
    modulated = gamma_n * phase
    real_part = np.real(modulated)
    return np.sqrt(np.mean(real_part**2))  # RMSE

# ========================
# 5. Startwerte & Grenzen
# ========================
initial_guess = [
    0.03, 1/33, 0.0,     # b1, f1, phi1
    0.08, 1/137, 0.0,    # b2, f2, phi2
    0.04, 1/33300, 0.0,  # b3, f3, phi3
    0.0008, 0.5          # c, d
]

bounds = [
    (0, 0.2), (1/1000, 0.1), (-np.pi, np.pi),
    (0, 0.2), (1/1000, 0.1), (-np.pi, np.pi),
    (0, 0.2), (1/100000, 0.1), (-np.pi, np.pi),
    (0, 0.01), (0.1, 1.5)
]

# ========================
# 6. Optimierung starten
# ========================
result = minimize(objective, initial_guess, bounds=bounds, method='L-BFGS-B')

# Ergebnisparameter
opt_params = result.x
final_rmse = result.fun

print("Optimierte Parameter:")
for i, name in enumerate(['b1', 'f1', 'phi1', 'b2', 'f2', 'phi2', 'b3', 'f3', 'phi3', 'c', 'd']):
    print(f"{name} = {opt_params[i]:.8f}")
print(f"\nFinaler RMSE der Realteile: {final_rmse:.5f}")

# ========================
# 7. Plot der optimierten Lsung
# ========================
beta_opt = beta_model(n, opt_params)
real_part_opt = np.real(gamma_n * np.exp(1j * beta_opt))

plt.figure(figsize=(14, 6))
plt.plot(n, real_part_opt, label='Re[  exp(i_opt(n))]')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Optimiertes Beta(n)-Modell: EulerFreese-Realteil')
plt.xlabel('n')
plt.ylabel('Realteil')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy.fft import fft, fftfreq

# ========================
# 1. Parameterraum
# ========================
N = 100           # Diskretisierungspunkte
c = 10            # Bandgrenzenparameter
rho_vals = np.linspace(1.01, 1.5, 50)

# ========================
# 2. Prolate-Matrix (Slepian)  Nherung fr P
# ========================
def build_prolate_matrix(N, c):
    # Eigenproblem fr diskretisierte Prolate-Funktionen
    n = np.arange(N)
    d = ((N - 1 - 2 * n) ** 2) / 4
    e = np.full(N - 1, c * (N - 1 - np.arange(1, N)) * (np.arange(1, N)) / 2)
    return d, e

d, e = build_prolate_matrix(N, c)
eigvals, eigvecs = eigh_tridiagonal(d, e)

# ========================
# 3. Projektionsoperatoren
# ========================
# P  Projektion auf die ersten k Prolate-Eigenfunktionen
k = 5
P = eigvecs[:, -k:] @ eigvecs[:, -k:].T  # P: N x N Projektion
I = np.identity(N)

# ========================
# 4. Skalierungsoperator ()
# ========================
def theta_matrix(rho, N):
    # einfache diagonale Skalierung
    return np.diag(rho ** (-0.5) * np.power(rho, -np.arange(N) / N))

# ========================
# 5. Spurformel auswerten
# ========================
delta_vals = []
for rho in rho_vals:
    theta = theta_matrix(rho, N)
    delta_rho = np.trace(theta @ (I - P) @ P)
    delta_vals.append(delta_rho)

# ========================
# 6. Plotten
# ========================
plt.figure(figsize=(10, 5))
plt.plot(rho_vals, delta_vals, marker='o', label=r'$\delta(\rho)$ (diskret)')
plt.title('Diskrete Spurformel-Komponente () nach Connes')
plt.xlabel(r'$\rho$')
plt.ylabel(r'$\delta(\rho)$')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# --- Bibliotheken ---
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.fft import fft, fftfreq

# --- 1. Beta(n)-Struktur + Dirac-Matrix ---
N = 1000
n = np.arange(1, N + 1)
beta_vals = (
    0.003652 * n**(-1.336085)
    + 0.045 * np.cos(2 * np.pi * 0.0043 * n + 2.23)
    + 0.2   * np.cos(2 * np.pi * 0.00078 * n - 1.85)
    + 0.01  * n**0.76
)

def dirac_beta(beta):
    D = np.zeros((len(beta), len(beta)), dtype=np.complex128)
    for i in range(1, len(beta)-1):
        D[i, i-1] = -0.5
        D[i, i+1] = 0.5
    return 1j * np.pi * D

D_beta = dirac_beta(beta_vals)
eigvals = np.linalg.eigvalsh(D_beta)
eigvals = eigvals[np.abs(eigvals) > 1e-10]  # Filter: Nullwerte vermeiden

# --- 2. Zeta-Determinante berechnen ---
def zeta_D(s, eigenvalues):
    return np.sum(np.abs(eigenvalues) ** (-s))

def derivative_zeta_at_0(eigenvalues, h=1e-5):
    return -(zeta_D(h, eigenvalues) - zeta_D(-h, eigenvalues)) / (2*h)

zeta_logdet = derivative_zeta_at_0(eigvals)
print("Log Zeta-Determinante:", zeta_logdet)

# --- 3. Zhlfunktion N() und Fluktuationen ---
eig_sorted = np.sort(np.abs(eigvals))
lams = np.linspace(0.01, np.max(eig_sorted), 300)
N_lam = np.array([np.sum(eig_sorted <= l) for l in lams])

# Hauptwachstum (Weyl-Fit)
slope, intercept, *_ = linregress(lams, N_lam)
N_fit = slope * lams + intercept
delta_N = N_lam - N_fit

# --- Plots ---
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(lams, N_lam, label="N()")
plt.plot(lams, N_fit, '--', label=f"Weyl-Fit ~ ^{slope:.3f}")
plt.xlabel(""); plt.ylabel("N()")
plt.title("Zhlfunktion N()")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(lams, delta_N, label="N()")
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel(""); plt.ylabel("N()")
plt.title("Spektrale Fluktuation")
plt.legend()
plt.tight_layout()
plt.show()

# --- 4. Fourieranalyse der Fluktuationen ---
delta_fft = np.abs(fft(delta_N - np.mean(delta_N)))
freqs = fftfreq(len(delta_N), d=(lams[1] - lams[0]))

plt.figure(figsize=(10, 4))
plt.plot(freqs[:len(freqs)//2], delta_fft[:len(delta_fft)//2])
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourierspektrum von N()")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Parameter definieren ===
N = 1000                      # Anzahl der Punkte im Beta-Raum
eps = 1e-4                    # Diskretisierungsschritt fr zentrale Differenz
n = np.arange(2, N + 2)       # Start bei n=2 (um log(1)=0 zu vermeiden)
beta = np.log(n)              # Beta-Skala: log(n)
beta_centered = beta - np.mean(beta)

# === Zentrale Differenzmatrix auf Beta-Werten ===
D = np.zeros((N, N))
for i in range(1, N - 1):
    D[i, i - 1] = -1
    D[i, i + 1] = +1
D /= (2 * eps)

# === Dirac-Operator H_ = i _ ===
H_beta = 1j * eps * D

# === Dirac-Matrixstruktur auf 2N-dimensionalem Raum ===
H_dirac = np.block([
    [np.zeros((N, N)), H_beta],
    [H_beta, np.zeros((N, N))]
])

# === Eigenwerte berechnen ===
eigvals = np.linalg.eigvalsh(H_dirac)

# === Plot: Spektrum des Dirac-Operators ===
plt.figure(figsize=(8, 5))
plt.plot(np.sort(eigvals.real), ".", markersize=2)
plt.axhline(0, color="gray", linestyle="--")
plt.title("Eigenwerte des Dirac-Operators im Beta-Raum")
plt.xlabel("Index")
plt.ylabel("Eigenwert (reell)")
plt.grid(True)
plt.tight_layout()
plt.show()

# === Optional: Fourier-Spektrum der Diagonalstruktur von beta(n) ===
from scipy.fft import rfft, rfftfreq

fft_beta = np.abs(rfft(beta_centered))
freqs = rfftfreq(N, d=1)

plt.figure(figsize=(8, 4))
plt.plot(freqs, fft_beta)
plt.title("Fourier-Spektrum von (n)")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
