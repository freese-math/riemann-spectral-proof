```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
from scipy.fft import fft, fftfreq
from sklearn.preprocessing import StandardScaler
from scipy.stats import entropy

# === GPU Device ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Verwendetes Gerät: {device}")

# === Pfade ===
beta_path = '/content/drive/MyDrive/beta_skala_24032025_optimum.csv'
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'

# === Daten laden ===
beta = pd.read_csv(beta_path, header=None, skiprows=1).squeeze().astype(float).values
zeta = pd.read_csv(zeta_path, header=None).squeeze().astype(float).values[:len(beta)]
primes = pd.read_csv(prime_path, header=None)[0].astype(str)
primes = pd.to_numeric(primes, errors='coerce').dropna().astype(int).values[:len(beta)]
log_primes = np.log(primes)

# === Standardisierung ===
scaler = StandardScaler()
beta_std = scaler.fit_transform(beta.reshape(-1, 1)).flatten()
zeta_std = scaler.fit_transform(zeta.reshape(-1, 1)).flatten()
log_primes_std = scaler.fit_transform(log_primes.reshape(-1, 1)).flatten()

# === Fourier-Transformation ===
def compute_spectrum(data):
    N = len(data)
    spectrum = np.abs(fft(data))[:N // 2]
    freqs = fftfreq(N)[:N // 2]
    return freqs, spectrum

freqs, beta_spec = compute_spectrum(beta_std)
_, zeta_spec = compute_spectrum(zeta_std)
_, prime_spec = compute_spectrum(log_primes_std)

# === Spektrale Entropien + Varianzen ===
def spectral_entropy(x):
    p = np.abs(x)**2
    p = p / np.sum(p)
    return entropy(p)

print("\nSpektrale Entropien:")
print(f"Beta: {spectral_entropy(beta_spec):.5f} | Zeta: {spectral_entropy(zeta_spec):.5f} | log(Prime): {spectral_entropy(prime_spec):.5f}")

print("Varianzen der Spektren:")
print(f"Beta: {np.var(beta_spec):.5f} | Zeta: {np.var(zeta_spec):.5f} | log(Prime): {np.var(prime_spec):.5f}")

# === Plot: Fourier-Spektren (log-y-Achse + Differenzen) ===
plt.figure(figsize=(14, 7))
plt.plot(freqs, zeta_spec, label='Zeta-Nullstellen', color='green')
plt.plot(freqs, prime_spec, label='log(Primzahlen)', color='blue')
plt.plot(freqs, beta_spec, label='Beta-Skala (optimiert)', color='orange', linestyle='--')
plt.plot(freqs, np.abs(beta_spec - zeta_spec), 'r:', label='Beta - Zeta')
plt.plot(freqs, np.abs(beta_spec - prime_spec), 'm:', label='Beta - log(Prime)')
plt.yscale("log")
plt.title('Fourier-Spektren & Differenzen (standardisiert, log-Achse)')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Sliding Window Entropie ===
def sliding_entropy(data, win=1024, step=512):
    entropies = []
    indices = []
    for i in range(0, len(data) - win, step):
        window = data[i:i+win]
        _, spec = compute_spectrum(window)
        entropies.append(spectral_entropy(spec))
        indices.append(i + win // 2)
    return indices, entropies

plt.figure(figsize=(12, 4))
for series, label, color in zip([beta_std, zeta_std, log_primes_std],
                                ["Beta", "Zeta", "log(Prime)"], ["orange", "green", "blue"]):
    idx, ent = sliding_entropy(series, win=4096, step=1024)
    plt.plot(idx, ent, label=label, color=color)
plt.title("Sliding-Window-Spektralentropie (Fenster=4096)")
plt.xlabel("Index")
plt.ylabel("Entropie")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Heatmap: Entropie vs Varianz (Segmentanalyse) ===
def segment_stats(data, win=2048):
    entropies, variances = [], []
    for i in range(0, len(data) - win, win):
        seg = data[i:i+win]
        _, spec = compute_spectrum(seg)
        entropies.append(spectral_entropy(spec))
        variances.append(np.var(spec))
    return np.array(entropies), np.array(variances)

beta_e, beta_v = segment_stats(beta_std)
zeta_e, zeta_v = segment_stats(zeta_std)
prime_e, prime_v = segment_stats(log_primes_std)

plt.figure(figsize=(6, 6))
plt.scatter(beta_e, beta_v, label='Beta', alpha=0.5, color='orange')
plt.scatter(zeta_e, zeta_v, label='Zeta', alpha=0.5, color='green')
plt.scatter(prime_e, prime_v, label='log(Prime)', alpha=0.5, color='blue')
plt.title("Entropie vs. Varianz (segmentiert)")
plt.xlabel("Spektrale Entropie")
plt.ylabel("Varianz")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.sparse.linalg import eigsh
from sklearn.preprocessing import StandardScaler

# === Datenpfade ===
beta_path = '/content/drive/MyDrive/beta_skala_24032025_optimum.csv'
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt'

# === Parameter für Sliding Window ===
window_size = 4096  # Größe des Fensters
stride = 1024       # Schrittweite zwischen den Fenstern
num_eigenvalues = 5 # Anzahl der zu berechnenden kleinsten Eigenwerte

# === Gesamtdaten einlesen ===
beta_df = pd.read_csv(beta_path, header=None, skiprows=1)
beta_raw = pd.to_numeric(beta_df[0], errors='coerce').dropna().values

zeta_raw = pd.read_csv(zeta_path, header=None).squeeze().astype(float).values

prime_raw = pd.read_csv(prime_path, header=None)[0]
prime_vals = pd.to_numeric(prime_raw, errors='coerce').dropna().values
log_primes = np.log(prime_vals)

# === Sicherstellen, dass alle Datenreihen gleich lang sind ===
min_length = min(len(beta_raw), len(zeta_raw), len(log_primes))
beta_raw = beta_raw[:min_length]
zeta_raw = zeta_raw[:min_length]
log_primes = log_primes[:min_length]

# === Standardisierung vorbereiten ===
scaler = StandardScaler()

# === Funktion zur Erstellung des Laplace-Operators ===
def create_laplacian(n):
    diagonals = [np.ones(n), -2*np.ones(n), np.ones(n)]
    return diags(diagonals, offsets=[-1, 0, 1], format='csc')

# === Sliding Window Analyse ===
eigenvalues_list = []
positions = []

for start in range(0, min_length - window_size + 1, stride):
    end = start + window_size
    positions.append(start + window_size // 2)

    # Daten im aktuellen Fenster extrahieren
    beta_segment = beta_raw[start:end]
    zeta_segment = zeta_raw[start:end]
    logp_segment = log_primes[start:end]

    # Standardisierung
    beta_std = scaler.fit_transform(beta_segment.reshape(-1, 1)).flatten()
    zeta_std = scaler.fit_transform(zeta_segment.reshape(-1, 1)).flatten()
    logp_std = scaler.fit_transform(logp_segment.reshape(-1, 1)).flatten()

    # Laplace-Operator erstellen
    laplacian = create_laplacian(window_size)

    # Kleinste Eigenwerte berechnen
    eigenvalues = eigsh(laplacian, k=num_eigenvalues, which='SM', return_eigenvectors=False)
    eigenvalues_list.append(eigenvalues)

# === Eigenwerte in ein numpy-Array umwandeln ===
eigenvalues_array = np.array(eigenvalues_list)

# === Visualisierung ===
plt.figure(figsize=(14, 8))

# Eigenwerte plotten
for i in range(num_eigenvalues):
    plt.plot(positions, eigenvalues_array[:, i], label=f'Eigenwert {i+1}', alpha=0.7)

plt.title('Verlauf der kleinsten Eigenwerte über die Datenpositionen')
plt.xlabel('Datenposition (Mittelpunkt des Fensters)')
plt.ylabel('Eigenwert')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# === Bibliotheken ===
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from scipy.fft import fftfreq

# === GPU verwenden ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Verwendetes Gerät: {device}")

# === Datenpfade ===
beta_path = '/content/drive/MyDrive/beta_skala_24032025_optimum.csv'
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt'

# === Daten einlesen ===
beta_raw = pd.read_csv(beta_path, header=None, skiprows=1)[0].astype(float).values
zeta_raw = pd.read_csv(zeta_path, header=None).squeeze().astype(float).values
prime_raw = pd.read_csv(prime_path, header=None)[0]
prime_vals = pd.to_numeric(prime_raw, errors='coerce').dropna().astype(float).values
log_primes = np.log(prime_vals)

# === Standardisierung ===
scaler = StandardScaler()
beta_std = scaler.fit_transform(beta_raw.reshape(-1, 1)).flatten()
zeta_std = scaler.fit_transform(zeta_raw.reshape(-1, 1)).flatten()
log_primes_std = scaler.fit_transform(log_primes.reshape(-1, 1)).flatten()

# === FFT via PyTorch ===
def compute_fft_gpu(signal_np):
    signal = torch.tensor(signal_np, dtype=torch.float32, device=device)
    N = signal.shape[0]
    fft_result = torch.fft.fft(signal)
    freqs = fftfreq(N)
    return freqs, fft_result

# === Dimension angleichen ===
N = min(len(beta_std), len(zeta_std), len(log_primes_std))
beta_std = beta_std[:N]
zeta_std = zeta_std[:N]
log_primes_std = log_primes_std[:N]

# === FFT anwenden ===
freqs, fft_beta = compute_fft_gpu(beta_std)
amps = torch.abs(fft_beta)
phases = torch.angle(fft_beta)

# === Dominante Frequenzen extrahieren ===
threshold = 0.05 * torch.max(amps)
mask = amps > threshold

dominant_freqs = freqs[mask.cpu()]
dominant_amps = amps[mask]
dominant_phases = phases[mask]

# === MOS-Konstruktion ===
n = torch.arange(N, device=device).float()
kernel = torch.zeros_like(n)

for f, a, p in zip(dominant_freqs, dominant_amps, dominant_phases):
    kernel += a * torch.cos(2 * np.pi * f * n + p)

beta_mos = kernel.cpu().numpy()

# === Korrelation berechnen ===
corr_zeta = np.corrcoef(beta_mos, zeta_std)[0, 1]
corr_logp = np.corrcoef(beta_mos, log_primes_std)[0, 1]
print(f"Korrelation MOS vs Zeta: {corr_zeta:.6f}")
print(f"Korrelation MOS vs log(Prime): {corr_logp:.6f}")

# === Visualisierung ===
plt.figure(figsize=(14, 6))
plt.plot(beta_mos, label="Beta-Skala (MOS rekonstruiert)", color='orange')
plt.plot(zeta_std, label="Zeta-Nullstellen (standardisiert)", color='green', alpha=0.6)
plt.plot(log_primes_std, label="log(Primzahlen) (standardisiert)", color='blue', alpha=0.6)
plt.title("MOS-Rekonstruktion vs Zeta-Nullstellen & log(Primzahlen)")
plt.xlabel("Index")
plt.ylabel("Standardisierter Wert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Imports ===
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Verwendetes Gerät: {device}")

# === Daten einlesen ===
beta_path = '/content/drive/MyDrive/beta_skala_24032025_optimum.csv'
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt'

# === Beta-Skala ===
beta_df = pd.read_csv(beta_path, header=None, skiprows=1)
beta_raw = pd.to_numeric(beta_df[0], errors='coerce').dropna().values
beta_std = StandardScaler().fit_transform(beta_raw.reshape(-1, 1)).flatten()
beta_std = torch.tensor(beta_std, dtype=torch.float32, device=device)

# === Zeta-Nullstellen ===
zeta_zeros = pd.read_csv(zeta_path, header=None).squeeze().astype(float).values
zeta_zeros_std = StandardScaler().fit_transform(zeta_zeros.reshape(-1, 1)).flatten()

# === Primzahlen ===
prime_raw = pd.read_csv(prime_path, header=None, dtype=str)[0]
primes = pd.to_numeric(prime_raw, errors='coerce').dropna().astype(float).values
log_primes = np.log(primes)
log_primes_std = StandardScaler().fit_transform(log_primes.reshape(-1, 1)).flatten()
```

```python
# === MOS-Korrelationen ===
window_size = 4096
step_size = 512

beta_cpu = beta_std.cpu().numpy()
mos_zeta, mos_logp, indices = [], [], []

for start in range(0, len(beta_cpu) - window_size, step_size):
    end = start + window_size
    beta_win = beta_cpu[start:end]
    zeta_win = zeta_zeros_std[start:end]
    logp_win = log_primes_std[start:end]

    mos_zeta.append(np.corrcoef(beta_win, zeta_win)[0, 1])
    mos_logp.append(np.corrcoef(beta_win, logp_win)[0, 1])
    indices.append(start)

mos_zeta_t = torch.tensor(mos_zeta, dtype=torch.float32, device=device)
mos_logp_t = torch.tensor(mos_logp, dtype=torch.float32, device=device)
```

```python
import torch
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("GPU aktiviert:", device)

# === Datenpfade (bitte ggf. anpassen)
beta_path = "/content/drive/MyDrive/beta_skala_24032025_optimum.csv"
zeta_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
prime_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt"

# === Einlesen der Daten
beta_df = pd.read_csv(beta_path, header=None, skiprows=1)
beta_raw = pd.to_numeric(beta_df[0], errors="coerce").dropna().values

zeta_raw = pd.read_csv(zeta_path, header=None).squeeze().astype(float).values

prime_raw = pd.read_csv(prime_path, header=None, dtype=str)[0]
prime_vals = pd.to_numeric(prime_raw, errors="coerce").dropna().values
log_primes = np.log(prime_vals)

# === Standardisierung
scaler = StandardScaler()
beta_std = scaler.fit_transform(beta_raw.reshape(-1, 1)).flatten()
zeta_std = scaler.fit_transform(zeta_raw.reshape(-1, 1)).flatten()
logp_std = scaler.fit_transform(log_primes.reshape(-1, 1)).flatten()

# === Dummy Betti-Kurve (falls nicht geladen)
# z. B. aus Sliding-Minima/Maxima extrahiert – hier simuliert
window_size = 4096
step = 512
n_windows = (len(beta_std) - window_size) // step
betti_counts = []

for i in range(n_windows):
    start = i * step
    end = start + window_size
    segment = beta_std[start:end]
    maxima = np.sum((segment[1:-1] > segment[:-2]) & (segment[1:-1] > segment[2:]))
    minima = np.sum((segment[1:-1] < segment[:-2]) & (segment[1:-1] < segment[2:]))
    betti_counts.append(maxima + minima)

betti_counts = np.array(betti_counts)
```

```python
from scipy.signal import argrelextrema

def compute_betti_curve(series, window=50):
    series = np.array(series)
    maxima = argrelextrema(series, np.greater, order=window)[0]
    minima = argrelextrema(series, np.less, order=window)[0]
    betti_curve = np.zeros_like(series)
    betti_curve[maxima] = 1
    betti_curve[minima] = 1
    return np.convolve(betti_curve, np.ones(window, dtype=int), mode='same')

betti_approx = compute_betti_curve(beta_std, window=50)

# Speichern für spätere Wiederverwendung
betti_df = pd.DataFrame(betti_approx, columns=["Betti_Approx"])
betti_path = "/content/drive/MyDrive/betti_kurve.csv"
betti_df.to_csv(betti_path, index=False)

print("Betti-Kurve berechnet und gespeichert:", betti_path)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import coherence
from scipy.stats import zscore
from tqdm import tqdm

# Freese-Parameter aus deiner Analyse (Bezeichnungen & deltas)
labels = [
    "7/33300", "3/99900", "1/66600", "1/137", "1/129.4", "1/34", "1/33",
    "1/10", "1/5", "484960/1000000", "1/2"
]
deltas = [
    7/33300, 3/99900, 1/66600, 1/137, 1/129.4, 1/34, 1/33,
    1/10, 1/5, 484960/1000000, 1/2
]

# Daten laden (ersetzen durch deine Pfade)
beta_path = "/content/drive/MyDrive/beta_skala_24032025_optimum.csv"
zeta_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"

beta = pd.read_csv(beta_path).iloc[:, 0].astype(float).values
zeta = pd.read_csv(zeta_path).iloc[:, 0].astype(float).values

# Standardisieren
beta_std = zscore(beta)
zeta_std = zscore(zeta[:len(beta)])  # gleich lang

# MOS-Korrelation berechnen (Sliding-Fenster)
def sliding_corr(x, y, window=2000):
    corrs = []
    for i in range(0, len(x) - window, window):
        x_seg = x[i:i+window]
        y_seg = y[i:i+window]
        corr = np.corrcoef(x_seg, y_seg)[0, 1]
        corrs.append(corr)
    return np.array(corrs)

# Scanne alle Δ aus Freese-Reihe
results = []
for label, delta in tqdm(zip(labels, deltas), total=len(deltas), desc="Freese Δ-Scan"):
    # Frequenz-Modulation auf Beta-Skala
    t = np.arange(len(beta_std))
    freese_mod = np.sin(2 * np.pi * delta * t)

    # Moduliere Beta
    beta_mod = beta_std * freese_mod

    # Sliding-Korrelation gegen Zeta
    mos_corr = sliding_corr(beta_mod, zeta_std)

    # Stärke als ∇ der MOS-Korrelation
    mos_grad = np.gradient(mos_corr)
    mos_strength = np.linalg.norm(mos_grad)

    # Speichern
    results.append((label, delta, mos_strength))

# In DataFrame
df = pd.DataFrame(results, columns=["Label", "Delta", "MOS_Strukturwert"])
df_sorted = df.sort_values("MOS_Strukturwert")

# Plotten
plt.figure(figsize=(10, 5))
bars = plt.bar(df_sorted["Label"], df_sorted["MOS_Strukturwert"], color="skyblue")
plt.xticks(rotation=45)
plt.ylabel("MOS-Struktur (Gradienten-Norm)")
plt.title("Optimierung der MOS-Korrelation mit Freese-Parametern")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

# Parameter
N = 100  # Diskretisierungspunkte
L = 10.0  # Intervallgröße
x = np.linspace(0, L, N)

# Diskretisierung des Laplace-Operators (für Vergleich)
dx = x[1] - x[0]
kinetic = -1j / dx * np.ones(N-1)

# Beispielpotential V(x): Sigmoid + Oszillation
def V(x):
    A, B, C, w = 1.0, 5.0, 5.0, 3.0
    return A / (1 + np.exp(-B * (x - C))) + np.sin(w * x)

V_diag = V(x)

# Hamilton-Matrix: H = -i d/dx + V(x)
main_diag = V_diag
off_diag = kinetic

# Eigenwerte berechnen
eigvals, eigvecs = eigh_tridiagonal(main_diag.real, off_diag.real)

# Plot der ersten Eigenwerte
plt.plot(eigvals[:20], 'o-', label='Re(Eigenwerte)')
plt.xlabel("Index")
plt.ylabel("Eigenwert")
plt.title("Spektrum des Operators H")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.linalg import eigh_tridiagonal

# Definitionsbereich
N = 1000            # Diskretisierungspunkte
x_min, x_max = -10, 10
x = np.linspace(x_min, x_max, N)
dx = x[1] - x[0]

# Parameter deines Potenzials (angepasst an deine Skizze)
A = 1.0
B = 5.0
C = 0.0
D = 0.2
w = 3.0

# Potentialfunktion: sigmoid + sinus
V = A / (1 + np.exp(-B * (x - C))) + D * np.sin(w * x)

# Kinetischer Teil (Diskretisierung von -d²/dx²)
main_diag = 2.0 / dx**2 + V
off_diag = -1.0 / dx**2 * np.ones(N-1)

# Eigenwertproblem lösen (nur kleinste k Eigenwerte)
eigvals, eigvecs = eigh_tridiagonal(main_diag, off_diag, select='i', select_range=(0, 20))

# Ergebnisse plotten
plt.figure(figsize=(8, 6))
for n in range(5):
    plt.plot(x, eigvecs[:, n] + eigvals[n], label=f'n = {n}')
plt.plot(x, V, 'k--', label='Potential V(x)')
plt.title('Eigenfunktionen & Potential des Operators H')
plt.xlabel('x'); plt.ylabel('Energie')
plt.legend()
plt.grid(True)
plt.show()

# Spektrum anzeigen
print("Numerisch berechnete Eigenwerte (Reihe):")
for i, val in enumerate(eigvals):
    print(f"λ_{i} = {val:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags
from scipy.linalg import eigh_tridiagonal

# Diskretisierung
x_min, x_max, N = -10, 10, 1000
x = np.linspace(x_min, x_max, N)
dx = x[1] - x[0]

# Potenzial (Sigmoid + Sinus-Modulation)
A, B, c = 1.0, 1.0, 0.0
D, omega = 0.2, 2.0
V = A / (1 + np.exp(-B * (x - c))) + D * np.sin(omega * x)

# Kinetischer Anteil: -d²/dx² (Diskretisierung über 3-Punkte-Stencil)
kinetic_diag = np.full(N, -2.0)
kinetic_offdiag = np.full(N - 1, 1.0)
laplacian = diags([kinetic_offdiag, kinetic_diag, kinetic_offdiag], [-1, 0, 1]) / dx**2

# Hamilton-Operator: H = -d²/dx² + V(x)
H_matrix = (-laplacian).toarray() + np.diag(V)

# Eigenwerte berechnen
num_eigen = 20
eigvals, eigvecs = eigh_tridiagonal(np.diag(H_matrix), np.diag(H_matrix, k=1), select='i', select_range=(0, num_eigen - 1))

# Visualisierung
plt.figure(figsize=(8, 5))
for n in range(5):
    plt.plot(x, eigvecs[:, n] + eigvals[n], label=f'n = {n}')
plt.plot(x, V, 'k--', label='Potential V(x)')
plt.title("Eigenfunktionen & Potential des Operators H")
plt.xlabel("x")
plt.ylabel("Energie")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy.linalg import norm

# Eigenfunktionen als numerische Lösung
# (angenommen: psi_list enthält normierte Eigenfunktionen ψ₀, ψ₁, ..., ψ₄)
# Beispiel: psi_list = [ψ₀(x), ψ₁(x), ..., ψ₄(x)]
# Diese wurden bereits bei der Schrödingerlösung berechnet

def test_orthogonality(psi_list, x):
    n = len(psi_list)
    inner_products = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            ip = simps(psi_list[i] * psi_list[j], x)
            inner_products[i, j] = ip
    return inner_products

# Beispielhafte Ausgabe
inner_products = test_orthogonality(psi_list, x)
print("Inneres Produkt-Matrix (Orthogonalitätstest):")
print(np.round(inner_products, 4))
```
