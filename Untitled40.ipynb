```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 📌 Beta-Skala-Funktionsmodell (Fourier + sigmoidale Korrektur)
def beta_model(N, A, B, C, w, phi):
    return A * np.log(N + 1) + B * np.sin(w * np.log(N + 1) + phi) + C

# 📌 Daten laden
file_path = "/content/drive/MyDrive/beta_scale_operator.csv"
try:
    beta_data = pd.read_csv(file_path)
    N_values = beta_data.iloc[:, 0].values
    beta_values = beta_data.iloc[:, 1].values
    print("✅ Daten erfolgreich geladen!")
except FileNotFoundError:
    print("❌ Fehler: Datei nicht gefunden!")
    raise SystemExit

# 📌 Prüfen auf NaN oder Inf
if np.isnan(N_values).any() or np.isnan(beta_values).any():
    print("⚠️ NaN-Werte gefunden, Bereinigung wird durchgeführt...")
    mask = ~np.isnan(beta_values)
    N_values = N_values[mask]
    beta_values = beta_values[mask]

if np.isinf(N_values).any() or np.isinf(beta_values).any():
    print("⚠️ Inf-Werte gefunden, Bereinigung wird durchgeführt...")
    mask = ~np.isinf(beta_values)
    N_values = N_values[mask]
    beta_values = beta_values[mask]

# 📌 Optimierungsparameter setzen
initial_guess = [-0.5, 500000, 100000, 0.00001, 0]  # Startwerte
bounds_lower = [-1, 0, 0, 0, -np.pi]  # Untere Grenze
bounds_upper = [1, 1e6, 1e6, 1, np.pi]  # Obere Grenze

# 📌 Curve Fitting durchführen mit optimierten Einstellungen
try:
    popt, _ = curve_fit(
        beta_model, N_values, beta_values,
        p0=initial_guess,
        bounds=(bounds_lower, bounds_upper),
        maxfev=10000  # Erhöht für stabilere Optimierung
    )
    A_opt, B_opt, C_opt, w_opt, phi_opt = popt
    print(f"✅ Optimierte Parameter gefunden: A={A_opt}, B={B_opt}, C={C_opt}, w={w_opt}, phi={phi_opt}")
except RuntimeError as e:
    print("❌ Fehler: Optimale Parameter nicht gefunden!")
    print("Fehlermeldung:", str(e))
    raise SystemExit

# 📌 Berechnung der optimierten Beta-Skala
beta_approx = beta_model(N_values, A_opt, B_opt, C_opt, w_opt, phi_opt)

# 📌 Visualisierung des Vergleichs
plt.figure(figsize=(10, 5))
plt.plot(N_values, beta_values, label="Original Beta-Skala", color='blue', alpha=0.6)
plt.plot(N_values, beta_approx, label="Fourier-gestützte Approximation", linestyle='dashed', color='red')
plt.xlabel("N")
plt.ylabel("Beta-Werte")
plt.title("Vergleich: Beta-Skala vs. Fourier-gestützte Approximation")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Daten laden
beta_data = pd.read_csv("/content/drive/MyDrive/beta_scale_operator.csv")
N_values = beta_data.iloc[:, 0].values
beta_values = beta_data.iloc[:, 1].values

# Sigmoid-Modell
def beta_sigmoid_model(N, A, B, C, D, w, phi):
    return A / (1 + np.exp(-B * (N - C))) + D * np.sin(w * N + phi)

# Phasenmoduliertes Sigmoid-Modell
def beta_sigmoid_phase_model(N, A, B, C, D, w, phi, phase_shift):
    return A / (1 + np.exp(-B * (N - C))) + D * np.sin(w * N + phi + phase_shift * N)

# Fourier-Modell
def fourier_model(N, A, B, C, w, phi):
    return A * np.sin(B * N + C) + w * np.cos(phi * N)

# Anfangswerte für die Optimierung
initial_params_sigmoid = [230000, 1000, 1500000, 8494, 1e-4, -44]
initial_params_phase = [230000, 1000, 1500000, 8494, 1e-4, -44, 1e-6]
initial_params_fourier = [10000, 100000, 10000, 1e-4, 0]

# Parameter anpassen für Fourier-Modell mit bounds
bounds_fourier = ([0, 0, 0, -np.inf, -np.inf], [np.inf, np.inf, np.inf, np.inf, np.inf])

# Curve Fitting durchführen
popt_sigmoid, _ = curve_fit(beta_sigmoid_model, N_values, beta_values, p0=initial_params_sigmoid, maxfev=50000)
popt_phase, _ = curve_fit(beta_sigmoid_phase_model, N_values, beta_values, p0=initial_params_phase, maxfev=50000)
popt_fourier, _ = curve_fit(fourier_model, N_values, beta_values, p0=initial_params_fourier, bounds=bounds_fourier, maxfev=50000)

# Ergebnisse visualisieren
plt.figure(figsize=(12, 6))
plt.scatter(N_values, beta_values, label="Original Beta-Skala", color='gray', alpha=0.5)
plt.plot(N_values, beta_sigmoid_model(N_values, *popt_sigmoid), label="Sigmoid-Modell", linestyle='dashed')
plt.plot(N_values, beta_sigmoid_phase_model(N_values, *popt_phase), label="Phasenmoduliertes Sigmoid-Modell", linestyle='dotted')
plt.plot(N_values, fourier_model(N_values, *popt_fourier), label="Fourier-Modell", linestyle='solid')
plt.xlabel("N")
plt.ylabel("Beta-Skalen-Werte")
plt.title("Vergleich verschiedener Modellierungen der Beta-Skala")
plt.legend()
plt.grid()
plt.show()

# Ergebnisse ausgeben
print("✅ Optimierte Parameter für das sigmoidale Modell:")
print(f"A={popt_sigmoid[0]}, B={popt_sigmoid[1]}, C={popt_sigmoid[2]}, D={popt_sigmoid[3]}, w={popt_sigmoid[4]}, phi={popt_sigmoid[5]}")

print("✅ Optimierte Parameter für das phasenmodulierte Sigmoid-Modell:")
print(f"A={popt_phase[0]}, B={popt_phase[1]}, C={popt_phase[2]}, D={popt_phase[3]}, w={popt_phase[4]}, phi={popt_phase[5]}, phase_shift={popt_phase[6]}")

print("✅ Optimierte Parameter für das Fourier-Modell:")
print(f"A={popt_fourier[0]}, B={popt_fourier[1]}, C={popt_fourier[2]}, w={popt_fourier[3]}, phi={popt_fourier[4]}")
```
