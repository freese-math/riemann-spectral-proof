```python
# ðŸ“Œ Notwendige Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ðŸ“Œ Pfade zu den DatensÃ¤tzen (CSV-Dateien)
primzahlen_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv"
zeta_zeros_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"

# ðŸ“Œ Daten laden mit Fehlerbehandlung
primzahlen = pd.read_csv(primzahlen_path, header=None, dtype=str, low_memory=False).iloc[:, 0]
zeta_zeros = pd.read_csv(zeta_zeros_path, header=None, dtype=str, low_memory=False).iloc[:, 0]

# ðŸ”¹ Konvertiere Strings zu Float und entferne fehlerhafte Werte
primzahlen = pd.to_numeric(primzahlen, errors='coerce').dropna().astype(float).values
zeta_zeros = pd.to_numeric(zeta_zeros, errors='coerce').dropna().astype(float).values

# ðŸ“Œ Begrenzung auf exakt 2 Millionen Werte
N = min(len(primzahlen), len(zeta_zeros), 2_000_000)
primzahlen, zeta_zeros = primzahlen[:N], zeta_zeros[:N]

print(f"âœ… {N} Primzahlen geladen.")
print(f"âœ… {N} Zeta-Nullstellen geladen.")

# ðŸ“Œ Modell 1: Logarithmische Skalierung
def log_modell(n, a, b):
    return a - b * np.log(n)

# ðŸ“Œ Modell 2: Fibonacci-Skala (aus [278â€ source])
def fibonacci_modell(n, a, b, c):
    return a - b * np.log(n) + c / np.sqrt(n)

# ðŸ“Œ Modell 3: Euler-Freese-Korrektur (aus [279â€ source])
def euler_freese_modell(n, a, b, c):
    return a - b * np.log(n) + c * np.exp(-1 / (np.pi * np.log(n + 1e-10)))

# ðŸ“Œ Curve-Fitting fÃ¼r alle drei Modelle
params_log, _ = curve_fit(log_modell, np.arange(1, N+1), zeta_zeros)
params_fib, _ = curve_fit(fibonacci_modell, np.arange(1, N+1), zeta_zeros)
params_euler, _ = curve_fit(euler_freese_modell, np.arange(1, N+1), zeta_zeros)

# ðŸ“Œ Fehlerberechnung
error_log = np.mean((zeta_zeros - log_modell(np.arange(1, N+1), *params_log))**2)
error_fib = np.mean((zeta_zeros - fibonacci_modell(np.arange(1, N+1), *params_fib))**2)
error_euler = np.mean((zeta_zeros - euler_freese_modell(np.arange(1, N+1), *params_euler))**2)

# ðŸ“Œ Bestes Modell auswÃ¤hlen
model_errors = {"Log-Modell": error_log, "Fibonacci-Modell": error_fib, "Euler-Freese-Modell": error_euler}
best_model_name = min(model_errors, key=model_errors.get)

if best_model_name == "Log-Modell":
    best_model = log_modell
    best_params = params_log
elif best_model_name == "Fibonacci-Modell":
    best_model = fibonacci_modell
    best_params = params_fib
else:
    best_model = euler_freese_modell
    best_params = params_euler

print(f"âœ… Bestes Modell: {best_model_name} mit Fehler: {model_errors[best_model_name]:.5f}")

# ðŸ“Œ Visualisierung der Zeta-Nullstellen vs. bestes Modell
plt.figure(figsize=(7, 5))
plt.scatter(np.arange(1, N+1), zeta_zeros, color="blue", s=1, label="Zeta-Nullstellen")
plt.plot(np.arange(1, N+1), best_model(np.arange(1, N+1), *best_params), "r--", label=f"{best_model_name}")
plt.xlabel("Index N")
plt.ylabel("Werte")
plt.title("Vergleich: Zeta-Nullstellen vs. Bestes Modell")
plt.legend()
plt.grid()
plt.show()

# ðŸ“Œ Vergleich: Primzahlen vs. Zeta-Nullstellen
plt.figure(figsize=(7, 5))
plt.plot(primzahlen, zeta_zeros, "g.", markersize=1, alpha=0.5, label="Primzahlen vs. Zeta-Nullstellen")
plt.xlabel("Primzahlen")
plt.ylabel("Zeta-Nullstellen")
plt.title("Vergleich von Primzahlen und Zeta-Spektrum")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigvals

# ðŸ”¹ Definiere die Beta-Werte
beta_values = [0.484906, 0.320100, 0.288237]  # Gefundene Beta-Werte

# ðŸ”¹ Definiere den Dirac-Operator fÃ¼r Beta
def dirac_operator(beta, x_derivative):
    """ Erstellt eine 2x2 Dirac-artige Matrix mit Beta """
    return np.array([[0, beta + x_derivative], [beta - x_derivative, 0]])

# ðŸ”¹ Berechne Eigenwerte fÃ¼r verschiedene Beta-Werte
x_derivative = 0.1  # Ableitungs-NÃ¤herung fÃ¼r den Operator
eigenvalues = [eigvals(dirac_operator(beta, x_derivative)) for beta in beta_values]

# ðŸ”¹ Visualisierung
plt.figure(figsize=(6, 4))
for i, beta in enumerate(beta_values):
    plt.scatter([beta]*2, eigenvalues[i], label=f"Beta = {beta}")
plt.axhline(y=0, color="gray", linestyle="--", alpha=0.5)
plt.xlabel("Beta-Wert")
plt.ylabel("Eigenwerte von H_Beta")
plt.title("Dirac-artige Eigenwerte fÃ¼r Beta-Skala")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# ðŸ“‚ 1ï¸âƒ£ Datei laden
dateipfad = "/content/drive/MyDrive/beta_skala_perfekt.csv"
df = pd.read_csv(dateipfad, header=None)

# ðŸ· 2ï¸âƒ£ Daten extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values

# ðŸ“Š 3ï¸âƒ£ Lineare Regression fÃ¼r den verbleibenden Drift
x = np.arange(len(beta_skala))
slope, intercept, _, _, _ = linregress(x, beta_skala)

# ðŸ”„ 4ï¸âƒ£ Letzte Korrektur anwenden
beta_skala_korrigiert = beta_skala - (slope * x + intercept) + 1.0  # Normalisierung auf 1.0

# ðŸ“ 5ï¸âƒ£ Neue Datei speichern
df_neu = pd.DataFrame([zeta_nullstellen, beta_skala_korrigiert])
datei_neu = "/content/drive/MyDrive/beta_skala_final.csv"
df_neu.to_csv(datei_neu, index=False, header=False)

# ðŸ“Š 6ï¸âƒ£ Finale Visualisierung
plt.figure(figsize=(12, 4))
plt.plot(beta_skala_korrigiert, label="Beta-Skala (final)", color="blue")
plt.axhline(1, color="gray", linestyle="dashed", label="Zielwert 1.0")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Finale perfekte Beta-Skala")
plt.show()

print(f"\nâœ… Finale Datei gespeichert: {datei_neu}")
```

```python
# ðŸ“¦ Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ðŸ“‚ 1ï¸âƒ£ Pfad zur Datei mit perfekter Skala
pfad = "/content/drive/MyDrive/beta_skala_perfekt.csv"

# ðŸ“¥ 2ï¸âƒ£ Datei laden
df = pd.read_csv(pfad, header=None)
zeta_nullstellen = df.iloc[0].values
beta_skala = df.iloc[1].values

# âš ï¸ 3ï¸âƒ£ Fehlerreihe berechnen (Abweichung von idealem Wert 1.0)
fehler = beta_skala - 1.0

# ðŸ“ˆ 4ï¸âƒ£ Fourier-Analyse
n = len(fehler)
frequenzen = fftfreq(n, d=1)[:n // 2]               # Nur positive Frequenzen
spektrum = np.abs(fft(fehler))[:n // 2]             # Betrag der Fourier-Koeffizienten

# ðŸ” 5ï¸âƒ£ Dominante Frequenzen anzeigen (Top 5)
top_indices = np.argsort(spektrum)[-5:][::-1]
dominante_frequenzen = frequenzen[top_indices]
dominante_amplituden = spektrum[top_indices]

# ðŸ“Š 6ï¸âƒ£ Visualisierung
plt.figure(figsize=(12, 5))
plt.plot(frequenzen, spektrum, label="Fourier-Spektrum des Fehlers", color="darkred")
plt.scatter(dominante_frequenzen, dominante_amplituden, color="blue", zorder=5, label="Dominante Frequenzen")
plt.title("Fourier-Spektrum der Abweichung von Beta-Skala")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ðŸ§¾ 7ï¸âƒ£ Ergebnisse ausgeben
print("âœ… Mittelwert des Fehlers:", np.mean(np.abs(fehler)))
print("ðŸ” Dominante Frequenzen im Fehlerverlauf:", dominante_frequenzen)
print("ðŸ” Korrespondierende Amplituden:", dominante_amplituden)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ðŸ“‚ 1ï¸âƒ£ Datei laden
dateipfad = "/content/drive/MyDrive/beta_skala_perfekt.csv"
df = pd.read_csv(dateipfad, header=None)

# ðŸ· 2ï¸âƒ£ Daten extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values
x = np.arange(len(beta_skala))  # Index-Werte

# ðŸ” 3ï¸âƒ£ Dominante Frequenzen & Amplituden aus Fourier-Analyse
dominante_frequenzen = [0.00000000e+00, 4.99737138e-07, 9.99474277e-07, 1.49921141e-06, 1.99894855e-06]
amplituden = [611320.24849869, 181757.61560862, 91789.55598573, 61429.52940368, 46168.57423272]

# ðŸ”„ 4ï¸âƒ£ Fourier-Korrekturterm konstruieren
korrektur_term = np.zeros_like(beta_skala)
for f, A in zip(dominante_frequenzen, amplituden):
    korrektur_term += A * np.cos(2 * np.pi * f * x)  # Kosinus-Oszillation

# âž• 5ï¸âƒ£ Beta-Skala korrigieren
beta_skala_korrigiert = beta_skala - korrektur_term

# ðŸ“ 6ï¸âƒ£ Neue Datei speichern
df_korrigiert = pd.DataFrame([zeta_nullstellen, beta_skala_korrigiert])
datei_korrigiert = "/content/drive/MyDrive/beta_skala_fourier_korrigiert.csv"
df_korrigiert.to_csv(datei_korrigiert, index=False, header=False)

# ðŸ“Š 7ï¸âƒ£ Visualisierung der neuen Beta-Skala
plt.figure(figsize=(12, 4))
plt.plot(beta_skala, label="Beta-Skala (vor Korrektur)", color="red", alpha=0.6)
plt.plot(beta_skala_korrigiert, label="Beta-Skala (nach Fourier-Korrektur)", color="blue")
plt.axhline(1, color="gray", linestyle="dashed", label="Zielwert 1.0")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Beta-Skala vor und nach Fourier-Korrektur")
plt.show()

print(f"\nâœ… Korrigierte Datei gespeichert: {datei_korrigiert}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ðŸ“‚ 1ï¸âƒ£ Beta-Skala laden
dateipfad = "/content/drive/MyDrive/beta_skala_fourier_korrigiert_v2.csv"
df = pd.read_csv(dateipfad, header=None)

# ðŸ· 2ï¸âƒ£ Werte extrahieren
zeta_nullstellen = df.iloc[0, :].values  # Zeta-Nullstellen
beta_skala = df.iloc[1, :].values  # Korrigierte Beta-Skala

# ðŸ“Œ 3ï¸âƒ£ Definition der Freese-Reihe
def freese_reihe(x):
    term1 = (7**x) / (33 * 137)
    term2 = 1 / np.exp(np.pi * 0.484964 * x)  # Beta-Wert als Exponent
    term3 = 7/33300  # Gefundene dominante Frequenz
    term4 = 7/3  # Hypothetischer Faktor
    return term1 + term2 + term3 + term4

# ðŸ“Š 4ï¸âƒ£ Werte berechnen
beta_freese = freese_reihe(zeta_nullstellen)

# ðŸ”„ 5ï¸âƒ£ Fehleranalyse
fehler = beta_skala - beta_freese

# ðŸ” 6ï¸âƒ£ Fourier-Analyse des Fehlers
N = len(fehler)
fft_spektrum = np.abs(fft(fehler))[:N//2]
frequenzen = fftfreq(N)[:N//2]

# ðŸ“‰ 7ï¸âƒ£ Plots
plt.figure(figsize=(12, 4))
plt.plot(zeta_nullstellen, beta_skala, label="Beta-Skala (Fourier-korrigiert)", color="red", alpha=0.6)
plt.plot(zeta_nullstellen, beta_freese, label="Freese-Reihe", color="blue", linestyle="dashed")
plt.xlabel("Index")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Beta-Skala vs. Freese-Reihe")
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(frequenzen, fft_spektrum, label="Fourier-Spektrum des Fehlers", color="red")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Abweichung von Freese-Reihe")
plt.legend()
plt.show()

# âœ… Ergebnis ausgeben
print(f"âœ… Mittelwert des Fehlers: {np.mean(fehler)}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ðŸ“‚ 1ï¸âƒ£ Datenquelle definieren
dateipfad = "/content/drive/MyDrive/beta_skala_fourier_korrigiert_v2.csv"
df = pd.read_csv(dateipfad, header=None)

# ðŸ“Š 2ï¸âƒ£ Zeta-Nullstellen und Beta-Skala extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values

# ðŸ”¢ 3ï¸âƒ£ Freese-Reihe definieren (numerisch stabilisiert)
def freese_reihe(x):
    term1 = (7**x) / (33 * 137)  # Kontrolle Ã¼ber exponentielles Wachstum
    term2 = np.exp(-np.pi * 0.484964 * x)  # Stabilisierte Exponentialfunktion
    return term1 + term2

# ðŸ”¬ 4ï¸âƒ£ Berechnung der Freese-Reihe fÃ¼r unsere Skala
x_values = np.arange(len(beta_skala))
freese_values = freese_reihe(x_values)

# âš ï¸ Warnungen abfangen (Overflow-Handling)
freese_values = np.nan_to_num(freese_values, nan=0.0, posinf=1e308, neginf=-1e308)

# ðŸ“ˆ 5ï¸âƒ£ Plot: Vergleich Beta-Skala vs. Freese-Reihe
plt.figure(figsize=(12, 4))
plt.plot(beta_skala, label="Beta-Skala (Fourier-korrigiert)", color="red", alpha=0.7)
plt.plot(freese_values, label="Freese-Reihe", color="blue", linestyle="dashed")
plt.xlabel("Index")
plt.ylabel("Beta-Wert")
plt.title("Vergleich: Beta-Skala vs. Freese-Reihe")
plt.legend()
plt.show()

# âœ… 6ï¸âƒ£ Analyse des Fehlers
fehler = beta_skala - freese_values
mittelwert_fehler = np.mean(fehler)

# ðŸ“Š 7ï¸âƒ£ Fourier-Analyse des Fehlers
fft_fehler = np.fft.fft(fehler)
frequenzen = np.fft.fftfreq(len(fehler))
amplituden = np.abs(fft_fehler)

# ðŸ† 8ï¸âƒ£ Dominante Frequenzen identifizieren
dominante_frequenzen = frequenzen[np.argsort(amplituden)[-5:]]  # Top 5 Frequenzen
dominante_amplituden = np.sort(amplituden)[-5:]

# ðŸ“ˆ 9ï¸âƒ£ Fehler-Spektrum visualisieren
plt.figure(figsize=(12, 4))
plt.plot(frequenzen[:len(frequenzen)//2], amplituden[:len(frequenzen)//2], color="brown", label="Fourier-Spektrum des Fehlers")
plt.scatter(dominante_frequenzen, dominante_amplituden, color="blue", label="Dominante Frequenzen")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Abweichung von Freese-Reihe")
plt.legend()
plt.show()

# ðŸ“ 1ï¸âƒ£0ï¸âƒ£ Speicherung der Korrektur
df_korrigiert = pd.DataFrame([zeta_nullstellen, freese_values])
datei_korrigiert = "/content/drive/MyDrive/beta_skala_freese_korrigiert.csv"
df_korrigiert.to_csv(datei_korrigiert, index=False, header=False)

# ðŸ“ 1ï¸âƒ£1ï¸âƒ£ Ausgabe der Analyse
print(f"âœ… Mittelwert des Fehlers: {mittelwert_fehler}")
print(f"ðŸ” Dominante Frequenzen: {dominante_frequenzen}")
print(f"ðŸ” Korrespondierende Amplituden: {dominante_amplituden}")
print(f"âœ… Korrigierte Datei gespeichert: {datei_korrigiert}")
```
