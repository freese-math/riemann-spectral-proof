```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.optimize import curve_fit

# ✅ Pfad zur Datei (lokal oder manuell heruntergeladen)
file_path = "/content/drive/MyDrive/zeros6.txt" # Falls die Datei an einem anderen Ort liegt, Pfad anpassen

# ✅ Nullstellen aus Datei laden
nullstellen = np.loadtxt(file_path)

# ✅ Berechnung der Nullstellenabstände
abstaende = np.diff(nullstellen)

# ✅ Überprüfen der ersten 10 Nullstellenabstände
print("✅ Datei erfolgreich geladen!")
print("Erste 10 Nullstellenabstände:", abstaende[:10])

# ✅ Erweiterte Fibonacci-Freese-Skalierungsformel mit exponentieller Dämpfung
def fibonacci_freese_damped(N, A, beta, gamma, lambd):
    """
    Erweiterte Fibonacci-Freese-Formel mit zusätzlicher exponentieller Dämpfung.
    """
    phi = (1 + np.sqrt(5)) / 2  # Goldener Schnitt
    g_N = (1 / (phi * 2 * np.pi)) + (np.exp(-phi) / (np.pi**2)) + (np.log(N) / (phi**3 * np.pi**3))
    return (A * N**beta * np.exp(g_N)) + (gamma * np.exp(-lambd * N))

# ✅ Funktion zur Berechnung des Bestimmtheitsmaßes R^2
def calculate_r2(actual, predicted):
    residuals = actual - predicted
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((actual - np.mean(actual))**2)
    return 1 - (ss_res / ss_tot)

# ✅ Initiale Schätzwerte für die Parameter
initial_params_ff_damped = [1.5, -0.1, 1.0, 0.01]

# ✅ Anpassung der Fibonacci-Freese-Formel mit exponentieller Dämpfung
params_ff_damped, _ = curve_fit(fibonacci_freese_damped, np.arange(1, len(abstaende) + 1), abstaende, p0=initial_params_ff_damped)

# ✅ Berechnung der Modellvorhersagen
ff_damped_predictions = fibonacci_freese_damped(np.arange(1, len(abstaende) + 1), *params_ff_damped)

# ✅ Berechnung von R^2 für das erweiterte Fibonacci-Freese-Modell mit Dämpfung
r2_ff_damped = calculate_r2(abstaende, ff_damped_predictions)

# ✅ Optimierte Parameter extrahieren
A_opt, beta_opt, gamma_opt, lambd_opt = params_ff_damped

# ✅ Ergebnisse ausgeben
print(f"📊 Optimierte Parameter für Fibonacci-Freese-Skalierung mit Dämpfung:")
print(f"A = {A_opt:.6f}, Beta = {beta_opt:.6f}")
print(f"Gamma (Dämpfungsfaktor) = {gamma_opt:.6f}, Lambda (Zerfallsrate) = {lambd_opt:.6f}")
print(f"Bestimmtheitsmaß R^2 = {r2_ff_damped:.6f}")

# ✅ Visualisierung der neuen Modellanpassung
plt.figure(figsize=(10, 6))
plt.scatter(np.arange(1, len(abstaende) + 1), abstaende, color='black', label="Echte Nullstellenabstände", s=1)
plt.plot(np.arange(1, len(abstaende) + 1), ff_damped_predictions, 'r-', label="Fibonacci-Freese + Dämpfung")
plt.xlabel("Index der Nullstellen (N)")
plt.ylabel("Abstand der Nullstellen (L(N))")
plt.title("Fibonacci-Freese-Skalierung mit exponentieller Dämpfung")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ✅ 1. Google Drive einbinden und Datei laden
from google.colab import drive
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.optimize import curve_fit

# ✅ Google Drive verbinden
drive.mount('/content/drive')

# ✅ Pfad zur Datei in Google Drive
file_path = "/content/drive/MyDrive/zeros6.txt"

# ✅ Nullstellen aus Datei laden
nullstellen = np.loadtxt(file_path)

# ✅ Berechnung der Nullstellenabstände
abstaende = np.diff(nullstellen)

# ✅ Überprüfen der ersten 10 Nullstellenabstände
print("✅ Datei erfolgreich geladen!")
print("Erste 10 Nullstellenabstände:", abstaende[:10])

# ✅ Erweiterte Fibonacci-Freese-Skalierungsformel mit fraktaler Korrektur
def fibonacci_freese_fractal(N, A, beta, gamma, lambd, alpha):
    """
    Fibonacci-Freese-Skalierung mit exponentieller Dämpfung und fraktaler Korrektur.
    """
    phi_const = (1 + np.sqrt(5)) / 2  # Goldener Schnitt
    g_N = (1 / (phi_const * 2 * np.pi)) + (np.exp(-phi_const) / (np.pi**2)) + (np.log(N) / (phi_const**3 * np.pi**3))

    # Fraktale Korrekturfunktion: Skaleninvariante Modifikation des Potenzgesetzes
    h_N = (1 + alpha * np.sin(np.log(N)))

    return (A * N**beta * np.exp(g_N) * h_N) + (gamma * np.exp(-lambd * N))

# ✅ Funktion zur Berechnung des Bestimmtheitsmaßes R^2
def calculate_r2(actual, predicted):
    residuals = actual - predicted
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((actual - np.mean(actual))**2)
    return 1 - (ss_res / ss_tot)

# ✅ Initiale Schätzwerte für die Parameter
initial_params = [1.5, -0.1, 1.0, 0.01, 0.1]

# ✅ Nullstellen-Indizes für das Fitten des Modells
N_values = np.arange(1, len(abstaende) + 1)

# ✅ Anpassung der Fibonacci-Freese-Formel mit fraktaler Korrektur
params_opt, _ = curve_fit(fibonacci_freese_fractal, N_values, abstaende, p0=initial_params)

# ✅ Berechnung der Modellvorhersagen
ff_fractal_predictions = fibonacci_freese_fractal(N_values, *params_opt)

# ✅ Berechnung von R^2 für das erweiterte Fibonacci-Freese-Modell
r2_ff_fractal = calculate_r2(abstaende, ff_fractal_predictions)

# ✅ Optimierte Parameter extrahieren
A_opt, beta_opt, gamma_opt, lambd_opt, alpha_opt = params_opt

# ✅ Ergebnisse ausgeben
print(f"📊 Optimierte Parameter für Fibonacci-Freese-Skalierung mit fraktaler Korrektur:")
print(f"A = {A_opt:.6f}, Beta = {beta_opt:.6f}")
print(f"Gamma (Dämpfungsfaktor) = {gamma_opt:.6f}, Lambda (Zerfallsrate) = {lambd_opt:.6f}")
print(f"Alpha (Fraktale Korrektur) = {alpha_opt:.6f}")
print(f"Bestimmtheitsmaß R^2 = {r2_ff_fractal:.6f}")

# ✅ Visualisierung der neuen Modellanpassung
plt.figure(figsize=(10, 6))
plt.scatter(N_values, abstaende, color='black', label="Echte Nullstellenabstände", s=1)
plt.plot(N_values, ff_fractal_predictions, 'r-', label="Fibonacci-Freese + Fraktale Korrektur")
plt.xlabel("Index der Nullstellen (N)")
plt.ylabel("Abstand der Nullstellen (L(N))")
plt.title("Fibonacci-Freese-Skalierung mit fraktaler Korrektur")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt

# -------------------------------
# ✅ Funktion zur Generierung der GOE-Matrix (Sparse)
# -------------------------------
def generate_sparse_goe_matrix(N):
    """Erzeugt eine Sparse GOE-Matrix mit symmetrischen Zufallselementen"""
    np.random.seed(42)  # Reproduzierbarkeit
    diag = np.random.randn(N)  # Hauptdiagonale
    off_diag = np.random.randn(N-1) / np.sqrt(2)  # Nebendiagonale

    # Sparse-Speicherung der symmetrischen Matrix
    mat = sp.diags([off_diag, diag, off_diag], offsets=[-1, 0, 1], format="csr")
    return mat

# -------------------------------
# ✅ Funktion zur Berechnung der Eigenwert-Abstände
# -------------------------------
def compute_eigenvalue_gaps(matrix, num_eigenvalues=200):
    """Berechnet die kleinsten Eigenwerte einer Matrix und deren Abstände"""
    eigenvalues = spla.eigsh(matrix, k=num_eigenvalues, which="SA", return_eigenvectors=False)
    eigenvalues.sort()
    return np.diff(eigenvalues)  # Eigenwertabstände zurückgeben

# -------------------------------
# 🚀 Hauptprogramm
# -------------------------------
N = 5000  # Reduzierte Matrixgröße für bessere Performance

# 🔍 GOE-Matrix generieren (Sparse!)
print(f"🔬 Erzeuge Sparse GOE-Matrix mit N={N}...")
goe_matrix = generate_sparse_goe_matrix(N)

# 📊 Eigenwert-Abstände berechnen
print(f"📊 Berechne Eigenwert-Abstände für GOE (N={N})...")
goe_gaps = compute_eigenvalue_gaps(goe_matrix)

# 🔍 Visualisierung der Eigenwert-Abstände
plt.figure(figsize=(8,5))
plt.hist(goe_gaps, bins=50, density=True, alpha=0.7, color="blue", label="GOE Eigenwert-Abstände")
plt.xlabel("Eigenwert-Abstand")
plt.ylabel("Dichte")
plt.legend()
plt.title(f"Eigenwert-Abstände der GOE-Matrix (N={N})")
plt.grid(True)
plt.show()

print("✅ Berechnung abgeschlossen!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq

# 📂 Lade Nullstellen
file_path = "/content/drive/MyDrive/zeros6.txt"
print(f"📂 Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
print(f"✅ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# 🔍 Berechne Nullstellen-Abstände
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)

# 📌 **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * n**(-1/2) + B * n**(-1) + C

# 📌 **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * n**(-1/2) + B * n**(-1) + C + w * np.cos(n * w + phi)

# 🎯 **Optimierung der Parameter**
initial_guess_ffs = [2, 0.9, 2500]
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=initial_guess_ffs)

initial_guess_ffo = [params_ffs[0], params_ffs[1], params_ffs[2], 0.1, 0]
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=initial_guess_ffo)

# 🔍 **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))

# 📌 **Ausgabe der optimierten Parameter**
print(f"✅ Optimierte FFS-Parameter: A={params_ffs[0]:.6f}, B={params_ffs[1]:.6f}, C={params_ffs[2]:.6f}")
print(f"✅ Optimierte FFO-Parameter: A={params_ffo[0]:.6f}, B={params_ffo[1]:.6f}, C={params_ffo[2]:.6f}, w={params_ffo[3]:.6f}, φ={params_ffo[4]:.6f}")
print(f"⚡ FFS Fehler: {error_ffs:.6f}")
print(f"⚡ FFO Fehler: {error_ffo:.6f}")

# 📊 **Plot: Vergleich FFS & FFO mit echten Nullstellen-Abständen**
plt.figure(figsize=(10, 5))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abstände", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS-Fit", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO-Fit", color="red", linestyle="dotted")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.legend()
plt.title("Vergleich: FFS vs. FFO vs. Reale Nullstellen-Abstände")
plt.show()

# 🔍 **Fourier-Analyse: Identifikation dominanter Frequenzen**
fft_values = fft(nullstellen_abstaende)
frequencies = fftfreq(len(nullstellen_abstaende), d=1)

plt.figure(figsize=(10, 5))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_values[:len(fft_values)//2]), color="purple")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstände")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq
import scipy.optimize as opt

# 📂 **Dateipfad für Nullstellen**
file_path = "/content/drive/MyDrive/zeros6.txt"

# 📂 **Lade Nullstellen & berechne Abstände**
print(f"📂 Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
print(f"✅ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# 🔍 **FFT-Analyse zur Frequenzbestimmung**
print("🔍 Berechne dominierende Frequenz mit Fourier-Analyse...")
N = len(nullstellen_abstaende)
freqs = fftfreq(N, d=1)
fft_values = np.abs(fft(nullstellen_abstaende))

# **Finde die dominante Frequenz**
dominant_freq = freqs[np.argmax(fft_values[1:N//2])]  # Nur positive Frequenzen betrachten
w_auto = abs(dominant_freq)
print(f"✅ Automatisch ermittelte Frequenz für FFO: w = {w_auto:.6f}")

# 📌 **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * n**(-1/2) + B * n**(-1) + C

# 📌 **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * n**(-1/2) + B * n**(-1) + C + w * np.cos(n * w + phi)

# 🔍 **Parameteroptimierung**
n_values = np.arange(1, len(nullstellen_abstaende) + 1)

# ✅ **Optimierung der FFS-Parameter**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"✅ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# ✅ **Optimierung der FFO-Parameter (inkl. w aus FFT)**
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=[2, 0.9, 1, w_auto, 0])
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"✅ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, φ={phi_ffo:.6f}")

# 🔍 **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
print(f"⚡ FFS Fehler: {error_ffs:.6f}")
print(f"⚡ FFO Fehler: {error_ffo:.6f}")

# 📊 **Visualisierung der Ergebnisse**
plt.figure(figsize=(10, 5))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abstände", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS-Fit", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO-Fit (optimiert)", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.legend()
plt.show()
```
