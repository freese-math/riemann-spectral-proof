```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# 📌 Nullstellen aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"
zeros = np.loadtxt(file_path)

# 📌 Box-Counting-Methode zur Berechnung von D_f
def compute_fractal_dimension(data, scales=np.logspace(-3, 1, 50)):
    counts = []

    for scale in scales:
        bins = np.arange(min(data), max(data), scale)
        hist, _ = np.histogram(data, bins=bins)
        counts.append(np.count_nonzero(hist))

    log_scales = np.log(1 / scales)
    log_counts = np.log(counts)

    slope, intercept, r_value, p_value, std_err = linregress(log_scales, log_counts)
    return slope, log_scales, log_counts

# 📌 Berechnung der fraktalen Dimension
D_f, log_scales, log_counts = compute_fractal_dimension(zeros)

# 📊 Visualisierung der Regression
plt.figure(figsize=(8,5))
plt.plot(log_scales, log_counts, 'bo', label="Datenpunkte")
plt.plot(log_scales, D_f * log_scales + log_counts[0], 'r--', label=f"Fit: D_f = {D_f:.6f}")
plt.xlabel(r"$\log(1/\epsilon)$")
plt.ylabel(r"$\log(N(\epsilon))$")
plt.title("Bestimmung der fraktalen Dimension der Nullstellenverteilung")
plt.legend()
plt.show()

# 📌 Ausgabe des Werts für D_f
print(f"📌 Berechnete Fraktale Dimension D_f: {D_f:.6f}")

# 📌 Überprüfung des Zusammenhangs mit β
beta_theo = 1 - (1 / D_f)
print(f"📌 Berechnetes β aus D_f: {beta_theo:.6f}")
```

```python
# ✅ **Installation fehlender Pakete**
!pip install numpy scipy matplotlib statsmodels tqdm

# ✅ **Importe**
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
from tqdm import tqdm

# ✅ **Dateipfade**
primzahlen_pfad = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt"
nullstellen_pfad = "/content/drive/MyDrive/zeros6.txt"

# ✅ **1️⃣ Lade Daten**
def load_numbers(file_path, dtype=float):
    with open(file_path, "r") as f:
        return np.array([dtype(line.strip()) for line in f])

primes = load_numbers(primzahlen_pfad, dtype=int)
zeros = load_numbers(nullstellen_pfad, dtype=float)

print(f"✅ {len(primes)} Primzahlen geladen.")
print(f"✅ {len(zeros)} Nullstellen geladen.")

# ✅ **2️⃣ Berechnung der Abstände**
abstaende = np.diff(zeros)

# ✅ **3️⃣ Fibonacci-Freese-Formel (FFF)**
def FFF(N, alpha, beta, gamma):
    return alpha * N**beta + gamma

N_values = np.arange(1, len(zeros))  # Berechnung für alle Nullstellen

# ✅ **4️⃣ Methode A: Standard Least-Squares Fit (LSS)**
popt_lss, _ = opt.curve_fit(FFF, N_values, np.cumsum(abstaende), p0=[1.88, 0.916, 2488])

# ✅ **5️⃣ Methode B: Gewichtetes Least-Squares Fit (WLS)**
weights = 1 / (1 + np.cumsum(abstaende))  # Größere Werte weniger gewichtet
popt_wls, _ = opt.curve_fit(FFF, N_values, np.cumsum(abstaende), sigma=weights, p0=[1.88, 0.916, 2488])

# ✅ **6️⃣ Methode C: Monte-Carlo-Optimierung**
num_samples = 1000
mc_results = []

for _ in tqdm(range(num_samples), desc="Monte-Carlo Simulation"):
    noise = np.random.normal(scale=0.01, size=len(N_values))  # Zufällige Störung
    popt_mc, _ = opt.curve_fit(FFF, N_values, np.cumsum(abstaende) + noise, p0=[1.88, 0.916, 2488])
    mc_results.append(popt_mc)

mc_results = np.array(mc_results)
alpha_mc, beta_mc, gamma_mc = np.mean(mc_results, axis=0)

# ✅ **7️⃣ Vergleich der Methoden**
print("\n📌 **Vergleich der Fit-Methoden:**")
print(f"🔹 LSS-Fit: α = {popt_lss[0]:.6f}, β = {popt_lss[1]:.6f}, γ = {popt_lss[2]:.6f}")
print(f"🔹 WLS-Fit: α = {popt_wls[0]:.6f}, β = {popt_wls[1]:.6f}, γ = {popt_wls[2]:.6f}")
print(f"🔹 MC-Opt: α = {alpha_mc:.6f}, β = {beta_mc:.6f}, γ = {gamma_mc:.6f}")

# ✅ **8️⃣ Fehleranalyse**
errors = np.cumsum(abstaende) - FFF(N_values, *popt_lss)

print("\n📊 **Fehlerstatistik:**")
print(f"🔹 Durchschnittlicher Fehler: {np.mean(errors):.6f}")
print(f"🔹 Standardabweichung: {np.std(errors):.6f}")
print(f"🔹 Maximaler Fehler: {np.max(errors):.6f}")
print(f"🔹 Minimaler Fehler: {np.min(errors):.6f}")

# ✅ **9️⃣ Fehler-Histogramm**
plt.figure(figsize=(10,5))
plt.hist(errors, bins=100, color="blue", alpha=0.7)
plt.axvline(np.mean(errors), color="red", linestyle="dashed", label="Mittelwert")
plt.title("Histogramm der Fehler")
plt.xlabel("Fehler")
plt.ylabel("Häufigkeit")
plt.legend()
plt.show()

# ✅ **🔟 Fourier-Analyse der Fehler**
fft_errors = np.abs(np.fft.fft(errors))[:len(errors)//2]
frequenzen = np.fft.fftfreq(len(errors))[:len(errors)//2]

plt.figure(figsize=(10,5))
plt.plot(frequenzen, fft_errors, color="green")
plt.title("Fourier-Analyse der Fehler")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid()
plt.show()

# ✅ **1️⃣1️⃣ Autokorrelationsprüfung der Fehler**
acf = sm.tsa.acf(errors, nlags=100)

plt.figure(figsize=(10,5))
plt.stem(acf, use_line_collection=True)
plt.title("Autokorrelation der Fehler")
plt.xlabel("Lag")
plt.ylabel("Autokorrelation")
plt.grid()
plt.show()

# ✅ **1️⃣2️⃣ Bootstrap-Resampling**
bootstrap_samples = 500
bootstrap_params = []

for _ in tqdm(range(bootstrap_samples), desc="Bootstrap-Resampling"):
    indices = np.random.choice(len(N_values), len(N_values), replace=True)
    popt_boot, _ = opt.curve_fit(FFF, N_values[indices], np.cumsum(abstaende)[indices], p0=[1.88, 0.916, 2488])
    bootstrap_params.append(popt_boot)

bootstrap_params = np.array(bootstrap_params)
alpha_boot, beta_boot, gamma_boot = np.mean(bootstrap_params, axis=0)

print("\n📌 **Bootstrap-Resampling Ergebnis:**")
print(f"🔹 α = {alpha_boot:.6f}, β = {beta_boot:.6f}, γ = {gamma_boot:.6f}")

# ✅ **1️⃣3️⃣ Finale Plots: Vergleich der Fits**
plt.figure(figsize=(10,5))
plt.plot(N_values, np.cumsum(abstaende), label="Kumulative Nullstellen-Abstände", color="black")
plt.plot(N_values, FFF(N_values, *popt_lss), "r--", label="LSS-Fit")
plt.plot(N_values, FFF(N_values, *popt_wls), "g--", label="WLS-Fit")
plt.plot(N_values, FFF(N_values, alpha_mc, beta_mc, gamma_mc), "b--", label="MC-Fit")
plt.xlabel("N")
plt.ylabel("Kumulierte Abstände")
plt.title("Vergleich der Fits")
plt.legend()
plt.grid()
plt.show()
```

```python
# ✅ Installiere notwendige Pakete (falls nicht vorhanden)
!pip install numpy scipy sympy mpmath matplotlib pandas

# ✅ Importiere Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft
import scipy.stats as stats
from sympy import pi, E, ln
from scipy.signal import find_peaks
from scipy.stats import pearsonr

# 📌 **1️⃣ Datei-Pfade setzen** (Passe diese an, falls du andere Dateipfade nutzt)
pfad_nullstellen = "/content/drive/MyDrive/zeros6.txt"

# 📌 **2️⃣ Funktion zum Laden der Nullstellen**
def load_numbers(file_path):
    """ Lädt eine Datei mit Zahlen, eine pro Zeile. """
    with open(file_path, 'r') as f:
        return np.array([float(line.strip()) for line in f])

# 📌 **3️⃣ Nullstellen laden**
zeros = load_numbers(pfad_nullstellen)
N_values = np.arange(1, len(zeros) + 1)

# 📌 **4️⃣ Definition der Freese-Funktion (FF)**
def FF(N, alpha, beta, gamma):
    return alpha * N**beta + gamma

# 📌 **5️⃣ Curve-Fitting zur Parameterbestimmung**
y_values = np.cumsum(np.diff(zeros))

popt, _ = opt.curve_fit(FF, N_values[:-1], y_values, p0=[1.88, 0.916, 2488])
alpha_opt, beta_opt, gamma_opt = popt

print(f"📌 **Optimierte Parameter für FF:**\n🔹 α = {alpha_opt:.6f}\n🔹 β = {beta_opt:.6f}\n🔹 γ = {gamma_opt:.6f}")

# 📌 **6️⃣ Fehlerberechnung**
fehler = y_values - FF(N_values[:-1], alpha_opt, beta_opt, gamma_opt)

print(f"\n📊 **Fehlerstatistik:**\n🔹 Durchschnittlicher Fehler: {np.mean(fehler):.6f}\n🔹 Standardabweichung: {np.std(fehler):.6f}\n🔹 Maximaler Fehler: {np.max(fehler):.6f}\n🔹 Minimaler Fehler: {np.min(fehler):.6f}")

# 📌 **7️⃣ Speicherung der Ergebnisse als CSV**
df = pd.DataFrame({"N": N_values[:-1], "Cumsum_Zeros": y_values, "Fehler": fehler})
csv_path = "/content/freese_function_results.csv"
df.to_csv(csv_path, index=False)
print(f"✅ Datei gespeichert: {csv_path}")

# 📌 **8️⃣ Visualisierung der Ergebnisse**
plt.figure(figsize=(10,5))
plt.loglog(N_values[:-1], y_values, label="Kumulative Abstände", color="blue")
plt.loglog(N_values[:-1], FF(N_values[:-1], alpha_opt, beta_opt, gamma_opt), '--', label="FF-Fit", color="red")
plt.xlabel("N")
plt.ylabel("Σ Abstände")
plt.title("FF-Fit der kumulierten Nullstellen-Abstände")
plt.legend()
plt.grid()
plt.show()
```

```python
# ✅ Import der benötigten Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.optimize import curve_fit
import gudhi as gd

# ✅ Pfad zur Rainbow Table (RT) Datei in Google Drive
rt_filepath = "/content/drive/MyDrive/rainbowtable_10mio.txt"

# ✅ Laden der Nullstellen aus der Rainbow Table
try:
    rt_nullstellen = np.loadtxt(rt_filepath, dtype=np.complex128, max_rows=1000000) # Begrenzung für Vergleichbarkeit
    print(f"✅ {len(rt_nullstellen)} Nullstellen erfolgreich geladen!")
except Exception as e:
    print(f"⚠ Fehler beim Laden der Datei: {e}")
    rt_nullstellen = None

# ✅ Abstände zwischen den Nullstellen berechnen
if rt_nullstellen is not None:
    rt_abstaende = np.diff(np.imag(rt_nullstellen))

    # ✅ Berechnung der fraktalen Dimension (Box-Counting-Methode)
    def box_counting(data, scales):
        counts = []
        for scale in scales:
            hist, _ = np.histogram(data, bins=np.arange(min(data), max(data) + scale, scale))
            counts.append(np.count_nonzero(hist))
        return counts

    scales = np.logspace(0.1, 2, num=30)  # Logarithmisch verteilte Skalierungsgrößen
    counts = box_counting(rt_abstaende, scales)

    # ✅ Lineare Regression für die fraktale Dimension
    log_scales = np.log(scales)
    log_counts = np.log(counts)
    slope, _ = np.polyfit(log_scales, log_counts, 1)
    df_rt = abs(slope)  # Fraktale Dimension berechnet

    # ✅ Berechnung von β aus D_f
    beta_rt = 2 - df_rt

    # ✅ Fourier-Analyse der Nullstellen-Abstände
    fft_result = fft(rt_abstaende)
    frequencies = np.fft.fftfreq(len(rt_abstaende))

    # ✅ Signifikante Peaks in der Fourier-Analyse extrahieren
    threshold = 0.05 * max(abs(fft_result))
    significant_freqs = frequencies[abs(fft_result) > threshold]

    # ✅ Berechnung der Betti-Zahlen (Homologie)
    def compute_betti_numbers(data):
        rips = gd.RipsComplex(points=data, max_edge_length=2)
        simplex_tree = rips.create_simplex_tree(max_dimension=2)
        simplex_tree.compute_persistence()
        return simplex_tree.betti_numbers()

    data_2D = np.column_stack((np.real(rt_nullstellen[:1000]), np.imag(rt_nullstellen[:1000])))
    betti_numbers_rt = compute_betti_numbers(data_2D)

    # ✅ Ergebnisse speichern
    results = {
        "Fraktale Dimension D_f": [df_rt],
        "Beta-Wert β": [beta_rt],
        "Signifikante Frequenzen (Fourier)": [significant_freqs[:5]],  # Begrenzung auf 5 relevante Frequenzen
        "Betti-Zahlen": [betti_numbers_rt]
    }

    df_results = pd.DataFrame(results)

    # ✅ CSV speichern
    csv_filepath = "/content/drive/MyDrive/fractal_fourier_analysis_rt.csv"
    df_results.to_csv(csv_filepath, index=False)

    print(f"✅ Ergebnisse gespeichert: {csv_filepath}")

    # ✅ Anzeige der Ergebnisse
    print(df_results)

    # ✅ Visualisierung der Abstände als Histogramm
    plt.figure(figsize=(10,5))
    plt.hist(rt_abstaende, bins=100, color='blue', alpha=0.7)
    plt.title("Histogramm der Nullstellen-Abstände (Rainbow Table)")
    plt.xlabel("Abstand")
    plt.ylabel("Häufigkeit")
    plt.grid(True)
    plt.show()

    # ✅ Visualisierung der Fourier-Analyse
    plt.figure(figsize=(10,5))
    plt.plot(frequencies[:len(frequencies)//2], abs(fft_result[:len(fft_result)//2]), color='red')
    plt.title("Fourier-Analyse der Nullstellen-Abstände")
    plt.xlabel("Frequenz")
    plt.ylabel("Spektrale Amplitude")
    plt.grid(True)
    plt.show()

    print("✅ Analyse abgeschlossen!")
```

```python
# ✅ Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib sympy

# ✅ Import der Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import solve_ivp
from sympy import symbols, diff, sin, cos, log

# ✅ Pfad zur zeros6 Odlysko Datei (Google Drive)
rt_pfad = "/content/drive/MyDrive/zeros6.txt"

# ✅ 1️⃣ Lade die Nullstellen-Daten aus der RT
def load_numbers(file_path, max_values=2001051):
    with open(file_path, 'r') as f:
        return np.array([float(line.strip()) for line in f][:max_values])

zeros_rt = load_numbers(rt_pfad)

print(f"✅ {len(zeros_rt)} Nullstellen erfolgreich geladen!")

# ✅ 2️⃣ Definiere die Freese-Formel (FF)
def FF(N, A, beta, C, B, w, phi):
    return A * N**beta + C * np.log(N) + B * np.sin(w * N + phi)

# ✅ 3️⃣ Fit der FF an die Daten
N_values = np.arange(1, len(zeros_rt) + 1)
y_values = np.cumsum(np.diff(zeros_rt, prepend=zeros_rt[0]))

popt, _ = curve_fit(FF, N_values, y_values, p0=[1.88, 0.916, 2488, 0, 0.1, 0])
A_opt, beta_opt, C_opt, B_opt, w_opt, phi_opt = popt

print(f"📌 **Optimierte Parameter für FF:**\n🔹 A = {A_opt:.6f}\n🔹 β = {beta_opt:.6f}\n🔹 C = {C_opt:.6f}\n🔹 B = {B_opt:.6f}\n🔹 w = {w_opt:.6f}\n🔹 φ = {phi_opt:.6f}")

# ✅ 4️⃣ Erste und zweite Ableitung von FF
def FF_derivative(N, A, beta, C, B, w, phi):
    L1 = A * beta * N**(beta - 1) + C / N + B * w * np.cos(w * N + phi)
    L2 = A * beta * (beta - 1) * N**(beta - 2) - C / N**2 - B * w**2 * np.sin(w * N + phi)
    return L1, L2

# ✅ 5️⃣ Definiere die Differenzialgleichung
def diff_eq(N, L, A, beta, C, B, w, phi):
    L1, L2 = FF_derivative(N, A, beta, C, B, w, phi)
    return L2 + (L1 / N) - (L / N**2)

# ✅ 6️⃣ Numerische Lösung der Differenzialgleichung
def solve_ff_diff_eq(A, beta, C, B, w, phi, N_max):
    def ode_system(N, L):
        return [diff_eq(N, L[0], A, beta, C, B, w, phi)]

    sol = solve_ivp(ode_system, [1, N_max], [0], t_eval=np.arange(1, N_max + 1))
    return sol.t, sol.y[0]

N_solve, L_solve = solve_ff_diff_eq(A_opt, beta_opt, C_opt, B_opt, w_opt, phi_opt, len(N_values))

# ✅ 7️⃣ Vergleich der numerischen Lösung mit den echten Daten
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Echte FF-Daten", color="blue")
plt.plot(N_solve, L_solve, '--', label="Numerische Lösung der FF-DGL", color="red")
plt.xlabel("N")
plt.ylabel("Σ Nullstellen-Abstände")
plt.title("Vergleich der numerischen Lösung mit echten Daten")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.integrate import solve_ivp

# ✅ Pfad zur Rainbowtable (RT)
file_path = "/content/drive/MyDrive/zeros6.txt"

# ✅ Daten laden
data = np.loadtxt(file_path)
N_values = np.arange(1, len(data) + 1)
y_values = np.cumsum(data)

# ✅ Alternative Approximation: Riemannsche Näherung
def riemann_approx(N):
    return N / np.log(N)

# ✅ Definition der Freese-Funktion
def freese_function(N, A, β, C, D, w, φ):
    return A * (N**β) + C * np.sin(w * N + φ) + D

# ✅ Curve Fit mit optimierten Startwerten
p0 = [1.88, 0.92, 0.002, -3100, 0.08, -9000]
popt, pcov = curve_fit(freese_function, N_values, y_values, p0=p0, maxfev=10000)

# ✅ Extrahierte Parameter
A_opt, β_opt, C_opt, D_opt, w_opt, φ_opt = popt

# ✅ Berechnung der Residuen
residuals = y_values - freese_function(N_values, *popt)
rms_error = np.sqrt(np.mean(residuals**2))

# ✅ Ergebnisse ausgeben
print(f"🔹 A = {A_opt}")
print(f"🔹 β = {β_opt}")
print(f"🔹 C = {C_opt}")
print(f"🔹 D = {D_opt}")
print(f"🔹 w = {w_opt}")
print(f"🔹 φ = {φ_opt}")
print(f"📊 RMS-Fehler: {rms_error}")

# ✅ Speichern als CSV
df_results = pd.DataFrame({
    "N": N_values,
    "FF-Werte": freese_function(N_values, *popt),
    "Riemann-Näherung": riemann_approx(N_values),
    "Residuen": residuals
})
csv_path = "/content/drive/MyDrive/freese_function_results.csv"
df_results.to_csv(csv_path, index=False)
print(f"✅ Datei gespeichert: {csv_path}")

# ✅ Visualisierung der Ergebnisse
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Echte FF-Daten", color="blue")
plt.plot(N_values, freese_function(N_values, *popt), "--", label="Optimierte FF-Funktion", color="red")
plt.plot(N_values, riemann_approx(N_values), ":", label="Riemannsche Näherung", color="green")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("Σ Nullstellen-Abstände")
plt.legend()
plt.title("Vergleich der Freese-Funktion mit der Riemannschen Näherung")
plt.grid(True)
plt.show()
```

```python
# 📌 Installation der benötigten Pakete
!pip install numpy scipy matplotlib gudhi

# 📌 Imports
import numpy as np
import matplotlib.pyplot as plt
import gudhi as gd
from scipy.stats import ks_2samp
from scipy.fftpack import fft
from scipy.spatial.distance import pdist, squareform

# 📌 Pfade zu den Datensätzen
path_rt = "/content/drive/MyDrive/rainbowtable_10mio.txt"
path_odlyzko = "/content/drive/MyDrive/zeros6.txt"

# 📌 Laden der Daten
def load_data(filepath, max_values=None):
    data = np.loadtxt(filepath, dtype=np.float64)
    return data[:max_values] if max_values else data

# ✅ RT & Odlyzko (erste 2 Mio.)
zeros_rt = load_data(path_rt, max_values=2001051)
zeros_odlyzko = load_data(path_odlyzko, max_values=2001051)

# 📌 1️⃣ Vergleich der Nullstellen
plt.figure(figsize=(10, 5))
plt.plot(zeros_odlyzko[:1000], 'b.', label="Odlyzko (erste 1000)")
plt.plot(zeros_rt[:1000], 'r.', label="Rainbow Table (erste 1000)")
plt.xlabel("Index")
plt.ylabel("Nullstellenwert")
plt.legend()
plt.title("Vergleich der ersten 1000 Nullstellen")
plt.grid()
plt.show()

# 📌 2️⃣ Verteilung der Nullstellen-Abstände
abstände_odlyzko = np.diff(zeros_odlyzko)
abstände_rt = np.diff(zeros_rt)

plt.figure(figsize=(10, 5))
plt.hist(abstände_odlyzko, bins=100, alpha=0.6, label="Odlyzko", color="blue", density=True)
plt.hist(abstände_rt, bins=100, alpha=0.6, label="Rainbow Table", color="red", density=True)
plt.xlabel("Abstände zwischen Nullstellen")
plt.ylabel("Normierte Häufigkeit")
plt.legend()
plt.title("Vergleich der Nullstellen-Abstandsverteilung")
plt.show()

# 📌 3️⃣ Kolmogorov-Smirnov-Test (KS-Test)
ks_stat, p_value = ks_2samp(abstände_odlyzko, abstände_rt)
print(f"📌 KS-Test zwischen Odlyzko & RT: KS-Statistik={ks_stat:.6f}, p-Wert={p_value:.6f}")

# 📌 4️⃣ Fraktale Dimension berechnen
def fractal_dimension(points, k=10):
    dists = squareform(pdist(points.reshape(-1, 1)))
    r = np.linspace(dists.min(), dists.max(), num=100)
    counts = np.array([(dists < radius).sum() for radius in r])
    log_r = np.log(r[r > 0])
    log_counts = np.log(counts[r > 0])
    return np.polyfit(log_r, log_counts, 1)[0]

D_f_odlyzko = fractal_dimension(zeros_odlyzko)
D_f_rt = fractal_dimension(zeros_rt)

print(f"📌 Fraktale Dimension Odlyzko: {D_f_odlyzko:.6f}")
print(f"📌 Fraktale Dimension RT: {D_f_rt:.6f}")

# 📌 5️⃣ FFT-Analyse der Nullstellen
fft_odlyzko = np.abs(fft(zeros_odlyzko))[:500]
fft_rt = np.abs(fft(zeros_rt))[:500]

plt.figure(figsize=(10, 5))
plt.plot(fft_odlyzko, 'b', label="FFT Odlyzko")
plt.plot(fft_rt, 'r', label="FFT RT")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.legend()
plt.title("Fourier-Analyse der Nullstellen")
plt.show()

# 📌 6️⃣ Berechnung der Betti-Zahlen
def compute_betti_numbers(data):
    rips = gd.RipsComplex(points=data.reshape(-1, 1), max_edge_length=2)
    simplex_tree = rips.create_simplex_tree(max_dimension=2)
    simplex_tree.compute_persistence()
    return simplex_tree.betti_numbers()

betti_odlyzko = compute_betti_numbers(zeros_odlyzko[:1000])
betti_rt = compute_betti_numbers(zeros_rt[:1000])

print(f"📌 Betti-Zahlen für Odlyzko: {betti_odlyzko}")
print(f"📌 Betti-Zahlen für Rainbow Table: {betti_rt}")
```

```python
# 📌 Import der notwendigen Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ✅ Pfad zur Datei in Google Drive
datei_pfad = "/content/drive/MyDrive/zeros6.txt"  # Odlyzko-Nullstellen (2 Mio)
# datei_pfad = "/content/drive/MyDrive/rainbowtable_10mio.txt"  # Falls RT genutzt wird

# 📌 Daten laden
with open(datei_pfad, "r") as f:
    nullstellen = np.array([float(line.strip()) for line in f])

# ✅ Berechnung der kumulierten Nullstellen-Abstände
abstaende = np.diff(nullstellen)  # Differenzen der Nullstellen berechnen
summe_abstaende = np.cumsum(abstaende)  # Kumulierte Summe

# 📌 Funktionale Form für das Fitten
def freese_function(N, A, beta):
    return A * N**beta

# ✅ Fit der Daten zur Bestimmung von A und β
N_values = np.arange(1, len(summe_abstaende) + 1)  # Indexwerte
popt, pcov = curve_fit(freese_function, N_values, summe_abstaende, p0=[1, 1.9])

# 📌 Extrahierte Parameter
A_opt, beta_opt = popt

print(f"📌 Optimierte Parameter für FF:")
print(f"🔹 A = {A_opt:.6f}")
print(f"🔹 β = {beta_opt:.6f}")

# ✅ Vergleich von Modell und echten Daten
plt.figure(figsize=(10, 5))
plt.plot(N_values, summe_abstaende, label="Echte Daten", color="blue")
plt.plot(N_values, freese_function(N_values, *popt), label="Anpassung FF-Funktion", linestyle="dashed", color="red")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("Σ Nullstellen-Abstände")
plt.title("Vergleich der numerischen Anpassung mit echten Daten")
plt.legend()
plt.grid()
plt.show()

# ✅ Ergebnisse speichern
ergebnis_df = pd.DataFrame({"N": N_values, "Σ Nullstellen-Abstände": summe_abstaende, "FF-Modell": freese_function(N_values, *popt)})
ergebnis_df.to_csv("/content/drive/MyDrive/freese_function_fit.csv", index=False)

print("✅ Datei gespeichert: /content/drive/MyDrive/freese_function_fit.csv")
```

```python
# ✅ Benötigte Bibliotheken installieren und importieren
!pip install numpy scipy matplotlib gudhi

import numpy as np
import scipy.optimize as opt
import scipy.fftpack as fft
import matplotlib.pyplot as plt
import gudhi as gd

# ✅ Lade die echten Odlyzko-Nullstellen (2 Mio.)
file_path = "/content/drive/MyDrive/zeros6.txt"  # Anpassen, falls Pfad abweicht
zeros = np.loadtxt(file_path)

# ✅ Berechne Nullstellen-Abstände
abstaende = np.diff(zeros)

# ✅ Definiere die erweiterte Freese-Funktion mit Korrekturtermen
def freese_function(N, A, beta, B, C):
    return A * (N ** beta) * (np.log(N) ** B) + C

# ✅ Werte für N erzeugen
N_values = np.arange(1, len(abstaende) + 1)
y_values = np.cumsum(abstaende)

# ✅ Nichtlineare Anpassung der FF-Funktion
popt, _ = opt.curve_fit(freese_function, N_values, y_values, p0=[2.0, 0.91, 0.1, 0.0])
A_opt, beta_opt, B_opt, C_opt = popt

# ✅ Vergleich mit der Riemannschen Näherung
def riemann_approx(N):
    return N / np.log(N)

riemann_values = riemann_approx(N_values)

# ✅ Fourier-Analyse der Nullstellen-Abstände
fft_spectrum = np.abs(fft.fft(abstaende))
freqs = fft.fftfreq(len(abstaende))

# ✅ Signifikante Frequenzen extrahieren
significant_freqs = freqs[np.argsort(fft_spectrum)[-2:]]  # Top-2 Frequenzen

# ✅ Berechnung der fraktalen Dimension mit Gudhi
def compute_fractal_dimension(data, max_dimension=1):
    rips = gd.RipsComplex(points=np.expand_dims(data, axis=1), max_edge_length=2)
    simplex_tree = rips.create_simplex_tree(max_dimension=max_dimension)
    return simplex_tree.betti_numbers()

fractal_dimension = compute_fractal_dimension(abstaende)

# ✅ Ergebnisse ausgeben
print(f"📌 Optimierte Parameter für FF:")
print(f"🔹 A = {A_opt}")
print(f"🔹 β = {beta_opt}")
print(f"🔹 B = {B_opt}")
print(f"🔹 C = {C_opt}\n")

print(f"📌 Signifikante Frequenzen in der Fourier-Analyse: {significant_freqs}")
print(f"📌 Fraktale Dimension: {fractal_dimension}")

# ✅ Ergebnisse visualisieren
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Echte FF-Daten", color="blue")
plt.plot(N_values, freese_function(N_values, A_opt, beta_opt, B_opt, C_opt),
         linestyle="dashed", color="red", label="Optimierte FF-Funktion")
plt.plot(N_values, riemann_values, linestyle="dotted", color="green", label="Riemannsche Näherung")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("Σ Nullstellen-Abstände")
plt.legend()
plt.title("Vergleich der Freese-Funktion mit der Riemannschen Näherung")
plt.show()

# ✅ Speichern der Ergebnisse
output_file = "/content/drive/MyDrive/freese_function_odlyzko.csv"
np.savetxt(output_file, np.column_stack([N_values, y_values]), delimiter=",",
           header="N,Echte_Summe_Nullstellen_Abstände", comments="")
print(f"✅ Datei gespeichert: {output_file}")
```
