```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# ðŸ“Œ Nullstellen aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"
zeros = np.loadtxt(file_path)

# ðŸ“Œ Box-Counting-Methode zur Berechnung von D_f
def compute_fractal_dimension(data, scales=np.logspace(-3, 1, 50)):
    counts = []

    for scale in scales:
        bins = np.arange(min(data), max(data), scale)
        hist, _ = np.histogram(data, bins=bins)
        counts.append(np.count_nonzero(hist))

    log_scales = np.log(1 / scales)
    log_counts = np.log(counts)

    slope, intercept, r_value, p_value, std_err = linregress(log_scales, log_counts)
    return slope, log_scales, log_counts

# ðŸ“Œ Berechnung der fraktalen Dimension
D_f, log_scales, log_counts = compute_fractal_dimension(zeros)

# ðŸ“Š Visualisierung der Regression
plt.figure(figsize=(8,5))
plt.plot(log_scales, log_counts, 'bo', label="Datenpunkte")
plt.plot(log_scales, D_f * log_scales + log_counts[0], 'r--', label=f"Fit: D_f = {D_f:.6f}")
plt.xlabel(r"$\log(1/\epsilon)$")
plt.ylabel(r"$\log(N(\epsilon))$")
plt.title("Bestimmung der fraktalen Dimension der Nullstellenverteilung")
plt.legend()
plt.show()

# ðŸ“Œ Ausgabe des Werts fÃ¼r D_f
print(f"ðŸ“Œ Berechnete Fraktale Dimension D_f: {D_f:.6f}")

# ðŸ“Œ ÃœberprÃ¼fung des Zusammenhangs mit Î²
beta_theo = 1 - (1 / D_f)
print(f"ðŸ“Œ Berechnetes Î² aus D_f: {beta_theo:.6f}")
```

```python
# ðŸ“Œ **Notwendige Bibliotheken installieren**
!pip install numpy matplotlib scipy sympy mpmath

# ðŸ“Œ **Bibliotheken importieren**
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from sympy import prime
from mpmath import zetazero
import pandas as pd

# ðŸ“Œ **Dateipfade fÃ¼r Primzahlen & Nullstellen**
primzahlen_pfad = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt"
zeros_pfad = "/content/drive/MyDrive/zeros6.txt"

# ðŸ“Œ **1ï¸âƒ£ Lade Primzahlen**
def load_primes(file_path):
    with open(file_path, "r") as f:
        primes = np.array([int(line.strip()) for line in f])
    return primes

# ðŸ“Œ **2ï¸âƒ£ Lade Zeta-Nullstellen**
def load_zeros(file_path):
    with open(file_path, "r") as f:
        zeros = np.array([float(line.strip()) for line in f])
    return zeros

# **ðŸ“Œ Lade die Daten**
primes = load_primes(primzahlen_pfad)
zeros = load_zeros(zeros_pfad)

# **ðŸ“Œ Validierung der Datenmengen**
assert len(primes) == 2_001_052, f"âš ï¸ Primzahlenanzahl unerwartet: {len(primes)}"
assert len(zeros) == 2_001_052, f"âš ï¸ Nullstellenanzahl unerwartet: {len(zeros)}"

print("âœ… 2,001,052 Primzahlen geladen.")
print("âœ… 2,001,052 Nullstellen geladen.")

# ðŸ“Œ **3ï¸âƒ£ Berechnung der Nullstellen-AbstÃ¤nde**
abstaende = np.diff(zeros)

# ðŸ“Œ **4ï¸âƒ£ Fibonacci-Freese-Formel**
def FFF(n, alpha, beta, gamma):
    return alpha * n ** beta + gamma

# ðŸ“Œ **5ï¸âƒ£ Curve-Fitting zur Optimierung von Î±, Î², Î³**
N_values = np.arange(1, len(zeros))
y_values = np.cumsum(abstaende)  # Kumulative AbstÃ¤nde

popt_fff, _ = opt.curve_fit(FFF, N_values, y_values, p0=[1.88, 0.916, 2488])
alpha_opt, beta_opt, gamma_opt = popt_fff

# ðŸ“Œ **6ï¸âƒ£ Fehlerstatistik berechnen**
fehler = y_values - FFF(N_values, alpha_opt, beta_opt, gamma_opt)
durchschnittlicher_fehler = np.mean(fehler)
std_fehler = np.std(fehler)
max_fehler = np.max(np.abs(fehler))

# ðŸ“Œ **7ï¸âƒ£ Ergebnisse ausgeben**
print("\nðŸ“Œ Optimierte Parameter fÃ¼r FFF:")
print(f"ðŸ”¹ Î± = {alpha_opt:.6f}")
print(f"ðŸ”¹ Î² = {beta_opt:.6f}")
print(f"ðŸ”¹ Î³ = {gamma_opt:.6f}")

print("\nðŸ“Š Fehlerstatistik:")
print(f"ðŸ”¹ Durchschnittlicher Fehler: {durchschnittlicher_fehler:.6f}")
print(f"ðŸ”¹ Standardabweichung: {std_fehler:.6f}")
print(f"ðŸ”¹ Maximaler Fehler: {max_fehler:.6f}")

# ðŸ“Œ **8ï¸âƒ£ Visualisierung: FFF vs. Nullstellen**
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Summierte Nullstellen-AbstÃ¤nde", color="blue")
plt.plot(N_values, FFF(N_values, alpha_opt, beta_opt, gamma_opt), label="FFF Fit", linestyle="dashed", color="red")
plt.xlabel("N")
plt.ylabel("Kumulative AbstÃ¤nde")
plt.title("Fibonacci-Freese-Formel Fit vs. Nullstellen-AbstÃ¤nde")
plt.legend()
plt.grid()
plt.show()

# ðŸ“Œ **9ï¸âƒ£ Korrelation zwischen Primzahlen & Nullstellen**
plt.figure(figsize=(10,5))
plt.scatter(primes[:len(zeros)], zeros, c='red', s=1, label="Primzahlen vs. Zeta-Nullstellen")
plt.xlabel("Primzahlen")
plt.ylabel("Zeta-Nullstellen")
plt.title("Korrelation: Primzahlen vs. Zeta-Nullstellen")
plt.legend()
plt.grid()
plt.show()
```

```python
# ðŸ“Œ Installation der benÃ¶tigten Pakete
!pip install numpy scipy matplotlib gudhi

# ðŸ“Œ Imports
import numpy as np
import matplotlib.pyplot as plt
import gudhi as gd
from scipy.stats import ks_2samp
from scipy.fftpack import fft
from scipy.spatial.distance import pdist, squareform

# ðŸ“Œ Pfade zu den DatensÃ¤tzen
path_rt = "/content/drive/MyDrive/rainbowtable_10mio.txt"
path_odlyzko = "/content/drive/MyDrive/zeros6.txt"

# ðŸ“Œ Laden der Daten
def load_data(filepath, max_values=None):
    data = np.loadtxt(filepath, dtype=np.float64)
    return data[:max_values] if max_values else data

# âœ… RT & Odlyzko (erste 2 Mio.)
zeros_rt = load_data(path_rt, max_values=2001051)
zeros_odlyzko = load_data(path_odlyzko, max_values=2001051)

# ðŸ“Œ 1ï¸âƒ£ Vergleich der Nullstellen
plt.figure(figsize=(10, 5))
plt.plot(zeros_odlyzko[:1000], 'b.', label="Odlyzko (erste 1000)")
plt.plot(zeros_rt[:1000], 'r.', label="Rainbow Table (erste 1000)")
plt.xlabel("Index")
plt.ylabel("Nullstellenwert")
plt.legend()
plt.title("Vergleich der ersten 1000 Nullstellen")
plt.grid()
plt.show()

# ðŸ“Œ 2ï¸âƒ£ Verteilung der Nullstellen-AbstÃ¤nde
abstÃ¤nde_odlyzko = np.diff(zeros_odlyzko)
abstÃ¤nde_rt = np.diff(zeros_rt)

plt.figure(figsize=(10, 5))
plt.hist(abstÃ¤nde_odlyzko, bins=100, alpha=0.6, label="Odlyzko", color="blue", density=True)
plt.hist(abstÃ¤nde_rt, bins=100, alpha=0.6, label="Rainbow Table", color="red", density=True)
plt.xlabel("AbstÃ¤nde zwischen Nullstellen")
plt.ylabel("Normierte HÃ¤ufigkeit")
plt.legend()
plt.title("Vergleich der Nullstellen-Abstandsverteilung")
plt.show()

# ðŸ“Œ 3ï¸âƒ£ Kolmogorov-Smirnov-Test (KS-Test)
ks_stat, p_value = ks_2samp(abstÃ¤nde_odlyzko, abstÃ¤nde_rt)
print(f"ðŸ“Œ KS-Test zwischen Odlyzko & RT: KS-Statistik={ks_stat:.6f}, p-Wert={p_value:.6f}")

# ðŸ“Œ 4ï¸âƒ£ Fraktale Dimension berechnen
def fractal_dimension(points, k=10):
    dists = squareform(pdist(points.reshape(-1, 1)))
    r = np.linspace(dists.min(), dists.max(), num=100)
    counts = np.array([(dists < radius).sum() for radius in r])
    log_r = np.log(r[r > 0])
    log_counts = np.log(counts[r > 0])
    return np.polyfit(log_r, log_counts, 1)[0]

D_f_odlyzko = fractal_dimension(zeros_odlyzko)
D_f_rt = fractal_dimension(zeros_rt)

print(f"ðŸ“Œ Fraktale Dimension Odlyzko: {D_f_odlyzko:.6f}")
print(f"ðŸ“Œ Fraktale Dimension RT: {D_f_rt:.6f}")

# ðŸ“Œ 5ï¸âƒ£ FFT-Analyse der Nullstellen
fft_odlyzko = np.abs(fft(zeros_odlyzko))[:500]
fft_rt = np.abs(fft(zeros_rt))[:500]

plt.figure(figsize=(10, 5))
plt.plot(fft_odlyzko, 'b', label="FFT Odlyzko")
plt.plot(fft_rt, 'r', label="FFT RT")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.legend()
plt.title("Fourier-Analyse der Nullstellen")
plt.show()

# ðŸ“Œ 6ï¸âƒ£ Berechnung der Betti-Zahlen
def compute_betti_numbers(data):
    rips = gd.RipsComplex(points=data.reshape(-1, 1), max_edge_length=2)
    simplex_tree = rips.create_simplex_tree(max_dimension=2)
    simplex_tree.compute_persistence()
    return simplex_tree.betti_numbers()

betti_odlyzko = compute_betti_numbers(zeros_odlyzko[:1000])
betti_rt = compute_betti_numbers(zeros_rt[:1000])

print(f"ðŸ“Œ Betti-Zahlen fÃ¼r Odlyzko: {betti_odlyzko}")
print(f"ðŸ“Œ Betti-Zahlen fÃ¼r Rainbow Table: {betti_rt}")
```

```python
# ðŸ“Œ Import der notwendigen Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# âœ… Pfad zur Datei in Google Drive
datei_pfad = "/content/drive/MyDrive/zeros6.txt"  # Odlyzko-Nullstellen (2 Mio)
# datei_pfad = "/content/drive/MyDrive/rainbowtable_10mio.txt"  # Falls RT genutzt wird

# ðŸ“Œ Daten laden
with open(datei_pfad, "r") as f:
    nullstellen = np.array([float(line.strip()) for line in f])

# âœ… Berechnung der kumulierten Nullstellen-AbstÃ¤nde
abstaende = np.diff(nullstellen)  # Differenzen der Nullstellen berechnen
summe_abstaende = np.cumsum(abstaende)  # Kumulierte Summe

# ðŸ“Œ Funktionale Form fÃ¼r das Fitten
def freese_function(N, A, beta):
    return A * N**beta

# âœ… Fit der Daten zur Bestimmung von A und Î²
N_values = np.arange(1, len(summe_abstaende) + 1)  # Indexwerte
popt, pcov = curve_fit(freese_function, N_values, summe_abstaende, p0=[1, 1.9])

# ðŸ“Œ Extrahierte Parameter
A_opt, beta_opt = popt

print(f"ðŸ“Œ Optimierte Parameter fÃ¼r FF:")
print(f"ðŸ”¹ A = {A_opt:.6f}")
print(f"ðŸ”¹ Î² = {beta_opt:.6f}")

# âœ… Vergleich von Modell und echten Daten
plt.figure(figsize=(10, 5))
plt.plot(N_values, summe_abstaende, label="Echte Daten", color="blue")
plt.plot(N_values, freese_function(N_values, *popt), label="Anpassung FF-Funktion", linestyle="dashed", color="red")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("Î£ Nullstellen-AbstÃ¤nde")
plt.title("Vergleich der numerischen Anpassung mit echten Daten")
plt.legend()
plt.grid()
plt.show()

# âœ… Ergebnisse speichern
ergebnis_df = pd.DataFrame({"N": N_values, "Î£ Nullstellen-AbstÃ¤nde": summe_abstaende, "FF-Modell": freese_function(N_values, *popt)})
ergebnis_df.to_csv("/content/drive/MyDrive/freese_function_fit.csv", index=False)

print("âœ… Datei gespeichert: /content/drive/MyDrive/freese_function_fit.csv")
```
