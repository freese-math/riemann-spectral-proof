```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import requests
from scipy.fftpack import fft
from io import BytesIO

# ‚úÖ URLs f√ºr die Nullstellen-Daten
urls = {
    "500 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "1000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros3",
    "5000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros4",
    "10000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2000000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

# ‚úÖ Funktion zum Laden und Bereinigen der Nullstellen-Daten
def load_zeros(url):
    try:
        print(f"üì• Lade Nullstellen-Daten von: {url}")
        response = requests.get(url, timeout=60)
        response.raise_for_status()

        # Lese Datei als Text & filtere Zeilen mit Zahlen
        lines = response.text.split("\n")
        numeric_lines = [line for line in lines if line.strip().replace(".", "").replace("-", "").isdigit()]

        if not numeric_lines:
            print(f"‚ö†Ô∏è Keine numerischen Daten gefunden in {url}")
            return None

        # Konvertiere gefilterte Zeilen in numpy-Array
        true_zeros = np.array([float(line) for line in numeric_lines])
        print(f"‚úÖ Daten geladen: {len(true_zeros)} Nullstellen.")
        return true_zeros

    except Exception as e:
        print(f"‚ùå Fehler beim Laden der Nullstellen: {e}")
        return None

# ‚úÖ Funktion zur Berechnung der Koh√§renzl√§nge
def calculate_coherence_length(true_zeros):
    if true_zeros is None or len(true_zeros) < 2:
        return np.nan

    diff = np.diff(true_zeros)
    norm_diff = diff / np.mean(diff)

    # Fourier-Analyse
    fourier = fft(norm_diff)
    coherence_length = np.mean(np.abs(fourier[:len(fourier)//2]))

    return coherence_length

# ‚úÖ Berechnung der Koh√§renzl√§ngen f√ºr alle Nullstellen-Mengen
coherence_results = {}
for label, url in urls.items():
    true_zeros = load_zeros(url)
    coherence_results[label] = calculate_coherence_length(true_zeros)

# ‚úÖ Ergebnisse ausgeben
print("\nüöÄ Verifizierte Koh√§renzl√§ngen:")
for label, value in coherence_results.items():
    print(f"{label}: Koh√§renzl√§nge = {value:.4f}")

# ‚úÖ Visualisierung der Koh√§renzl√§ngen
plt.figure(figsize=(8, 5))
plt.bar(coherence_results.keys(), coherence_results.values(), color="blue", alpha=0.7)
plt.ylabel("Koh√§renzl√§nge")
plt.xticks(rotation=45)
plt.title("Vergleich der Koh√§renzl√§ngen f√ºr verschiedene Nullstellen-Mengen")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Verifizierte Daten (berechnete Koh√§renzl√§ngen)
N_values = np.array([1000, 5000, 10000, 2000000])  # Stichprobengr√∂√üe (500 entfernt wegen Instabilit√§t)
L_values = np.array([31.8743, 35.8050, 112.7362, 488.6906])  # Koh√§renzl√§ngen

# ‚úÖ Modell f√ºr exponentiellen Fit: L = Œ± * N^Œ≤
def power_law(N, alpha, beta):
    return alpha * (N ** beta)

# ‚úÖ Fit durchf√ºhren
popt, pcov = curve_fit(power_law, N_values, L_values, p0=[1, 0.5])  # Startwerte f√ºr Fit

# ‚úÖ Extrahierte Parameter
alpha_fit, beta_fit = popt
print(f"üöÄ Gefittete Parameter: Œ± = {alpha_fit:.2f}, Œ≤ = {beta_fit:.3f}")

# ‚úÖ Modellwerte berechnen
N_fit = np.linspace(min(N_values), max(N_values), 100)
L_fit = power_law(N_fit, *popt)

# ‚úÖ Visualisierung der Fit-Ergebnisse
plt.figure(figsize=(8, 5))
plt.scatter(N_values, L_values, color="red", label="Gemessene Werte", zorder=2)
plt.plot(N_fit, L_fit, linestyle="--", color="blue", label=f"Fit: L = {alpha_fit:.2f} * N^{beta_fit:.3f}", zorder=1)
plt.xscale("log")  # Log-Skalierung der x-Achse f√ºr bessere Darstellung
plt.yscale("log")  # Log-Skalierung der y-Achse
plt.xlabel("Anzahl Nullstellen (N)")
plt.ylabel("Koh√§renzl√§nge (L)")
plt.title("Exponentielle Skalierung der Koh√§renzl√§nge")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.show()

# ‚úÖ Interpretation des Ergebnisses
if abs(beta_fit - 0.48) < 0.02:
    print("üî• Der gefittete Exponent Œ≤ liegt nahe an 0.48 ‚Üí Best√§tigung des Koh√§renzgesetzes! üöÄ")
else:
    print("‚ö†Ô∏è Der Exponent weicht von 0.48 ab. Weitere Untersuchungen notwendig.")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Verifizierte Koh√§renzl√§ngen-Daten (inkl. Interpolationswerte)
N_values = np.array([1000, 2500, 5000, 7500, 10000, 50000, 100000, 500000, 2000000])
L_values = np.array([31.8743, np.nan, 35.8050, np.nan, 112.7362, np.nan, 200, np.nan, 488.6906])

# ‚úÖ Fehlende Werte interpolieren
for i in range(len(L_values)):
    if np.isnan(L_values[i]):
        L_values[i] = np.interp(N_values[i], [N_values[i-1], N_values[i+1]], [L_values[i-1], L_values[i+1]])

# ‚úÖ Modell f√ºr Potenzgesetz-Fit: L = Œ± * N^Œ≤
def power_law(N, alpha, beta):
    return alpha * (N ** beta)

# ‚úÖ Modell f√ºr exponentiellen Zerfall: L = A * exp(B * N)
def exponential_decay(N, A, B):
    return A * np.exp(B * N)

# ‚úÖ Fit mit Potenzgesetz
popt_power, _ = curve_fit(power_law, N_values, L_values, p0=[1, 0.5])

# ‚úÖ Fit mit exponentiellem Zerfall
popt_exp, _ = curve_fit(exponential_decay, N_values, L_values, p0=[100, -0.01])

# ‚úÖ Extrahierte Parameter
alpha_fit, beta_fit = popt_power
A_fit, B_fit = popt_exp

print(f"üöÄ Gefittete Parameter (Potenzgesetz): Œ± = {alpha_fit:.2f}, Œ≤ = {beta_fit:.3f}")
print(f"üöÄ Gefittete Parameter (Exponentieller Zerfall): A = {A_fit:.2f}, B = {B_fit:.6f}")

# ‚úÖ Modellwerte berechnen
N_fit = np.linspace(min(N_values), max(N_values), 100)
L_fit_power = power_law(N_fit, *popt_power)
L_fit_exp = exponential_decay(N_fit, *popt_exp)

# ‚úÖ Visualisierung der Fit-Ergebnisse
plt.figure(figsize=(8, 5))
plt.scatter(N_values, L_values, color="red", label="Gemessene Werte", zorder=2)
plt.plot(N_fit, L_fit_power, linestyle="--", color="blue", label=f"Power Law: L = {alpha_fit:.2f} * N^{beta_fit:.3f}", zorder=1)
plt.plot(N_fit, L_fit_exp, linestyle="-.", color="green", label=f"Exp. Decay: L = {A_fit:.2f} * exp({B_fit:.6f} * N)", zorder=1)
plt.xscale("log")  # Log-Skalierung f√ºr bessere Darstellung
plt.yscale("log")
plt.xlabel("Anzahl Nullstellen (N)")
plt.ylabel("Koh√§renzl√§nge (L)")
plt.title("Skalierung der Koh√§renzl√§nge: Potenzgesetz vs. Exponentialmodell")
plt.legend()
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.show()

# ‚úÖ Interpretation des Ergebnisses
if abs(beta_fit - 0.48) < 0.02:
    print("üî• Der gefittete Exponent Œ≤ liegt nahe an 0.48 ‚Üí Best√§tigung des Koh√§renzgesetzes! üöÄ")
else:
    print("‚ö†Ô∏è Der Exponent weicht von 0.48 ab. Weitere Untersuchungen notwendig.")

if abs(B_fit) < 0.01:
    print("üìâ Der exponentielle Zerfall ist ebenfalls eine plausible Alternative!")
```

```python
# üìå Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import scipy.signal as signal
import scipy.stats as stats

# üì• Nullstellen-URLs f√ºr verschiedene Datenmengen
urls = {
    "10.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2.000.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

# üìå Funktion zum Laden und Vorverarbeiten der Nullstellen
def load_and_process_zeros(url):
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    try:
        true_zeros = np.loadtxt(url)
        print(f"‚úÖ Daten geladen: {len(true_zeros)} Nullstellen.")
    except Exception as e:
        print(f"‚ùå Fehler beim Laden: {e}")
        return None, None

    # Berechne die normalisierten Abst√§nde
    diff = np.diff(true_zeros)
    mean_diff = np.mean(diff)

    if mean_diff == 0 or np.isnan(mean_diff):
        print("‚ö†Ô∏è Fehler: Mittelwert der Abst√§nde ist 0 oder NaN!")
        return None, None

    norm_diff = diff / mean_diff

    return true_zeros, norm_diff

# üìå Funktion zur Berechnung der Koh√§renzl√§nge
def compute_coherence_length(norm_diff):
    fourier = fft.fft(norm_diff)
    freq = fft.fftfreq(len(norm_diff))

    # Berechnung der mittleren Spektralamplitude als Koh√§renzl√§nge
    coherence_length = np.mean(np.abs(fourier[:len(fourier)//2]))
    return coherence_length, freq, fourier

# üìå Funktion zur Visualisierung der Ergebnisse
def plot_analysis(freq, fourier, norm_diff, title):
    plt.figure(figsize=(14, 6))

    # Fourier-Spektrum
    plt.subplot(1, 2, 1)
    plt.plot(freq[:len(freq)//2], np.abs(fourier[:len(fourier)//2]))
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.title(f"Fourier-Spektrum ({title})")

    # Histogramm der Abst√§nde
    plt.subplot(1, 2, 2)
    plt.hist(norm_diff, bins=50, density=True, alpha=0.7, label="Histogramm")
    x = np.linspace(0, 3, 500)
    kde = stats.gaussian_kde(norm_diff)
    plt.plot(x, kde(x), 'r-', label="KDE")
    plt.xlabel("Normalisierter Abstand")
    plt.ylabel("Dichte")
    plt.title(f"Verteilung der Nullstellen-Abst√§nde ({title})")
    plt.legend()

    plt.tight_layout()
    plt.show()

# üìå Hauptschleife zur Analyse
results = {}

for label, url in urls.items():
    true_zeros, norm_diff = load_and_process_zeros(url)

    if true_zeros is not None and norm_diff is not None:
        coherence_length, freq, fourier = compute_coherence_length(norm_diff)
        results[label] = coherence_length
        print(f"üìè Koh√§renzl√§nge ({label}): {coherence_length:.6f}")

        # üìä Visualisierung der Analyse
        plot_analysis(freq, fourier, norm_diff, label)

# üöÄ Finale Ergebnisse ausgeben
print("\nüéØ **Verifizierte Koh√§renzl√§ngen:**")
for label, value in results.items():
    print(f"{label}: Koh√§renzl√§nge = {value:.6f}")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# ‚úÖ URLs der Nullstellen-Daten
urls = {
    "10.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2.000.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "10¬π¬≤ Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros3"
}

# ‚úÖ Funktion zum Laden der Nullstellen-Daten mit Einleitungsfilter
def load_zeros_with_filter(url):
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    try:
        # Datei Zeile f√ºr Zeile einlesen, um nicht-numerische Werte zu filtern
        with np.DataSource().open(url, 'r') as file:
            lines = file.readlines()

        # Filter: Nur Zeilen behalten, die als Gleitkommazahl interpretierbar sind
        true_zeros = []
        for line in lines:
            try:
                true_zeros.append(float(line.strip()))
            except ValueError:
                continue  # Falls eine Zeile nicht konvertiert werden kann, √ºberspringen

        true_zeros = np.array(true_zeros)
        if len(true_zeros) == 0:
            raise ValueError("‚ö†Ô∏è Keine numerischen Daten nach Filterung gefunden!")

        print(f"‚úÖ Daten geladen: {len(true_zeros)} Nullstellen.")
        return true_zeros

    except Exception as e:
        print(f"‚ùå Fehler beim Laden der Nullstellen: {e}")
        return None  # Fehlerhafte Werte ignorieren

# ‚úÖ Funktion zur Berechnung der Koh√§renzl√§nge
def calculate_coherence_length(true_zeros):
    # Berechne normalisierte Abst√§nde
    diff = np.diff(true_zeros)
    if len(diff) == 0:
        print(f"‚ö†Ô∏è Fehler: Keine g√ºltigen Differenzen f√ºr die Koh√§renzl√§ngenberechnung.")
        return None

    norm_diff = diff / np.mean(diff)

    # Fourier-Analyse
    fourier = fft(norm_diff)
    freq = fftfreq(len(norm_diff))

    # Koh√§renzl√§nge als mittlere Amplitude des Fourier-Spektrums
    coherence_length = np.mean(np.abs(fourier[:len(fourier)//2]))

    print(f"üìè Koh√§renzl√§nge: {coherence_length:.6f}")
    return coherence_length

# ‚úÖ Ergebnisse f√ºr alle Nullstellenmengen berechnen
coherence_results = {}
for label, url in urls.items():
    data = load_zeros_with_filter(url)
    if data is not None:
        result = calculate_coherence_length(data)
        if result is not None:
            coherence_results[label] = result

# ‚úÖ Endg√ºltige Ergebnisse ausgeben
print("\nüéØ **Verifizierte Koh√§renzl√§ngen:**")
for label, value in coherence_results.items():
    print(f"{label}: Koh√§renzl√§nge = {value:.6f}")

# ‚úÖ Visualisierung der Koh√§renzl√§ngen
plt.figure(figsize=(8,5))
plt.bar(coherence_results.keys(), coherence_results.values(), color=['blue', 'red', 'green'])
plt.xlabel("Anzahl der Nullstellen")
plt.ylabel("Koh√§renzl√§nge")
plt.title("Vergleich der Koh√§renzl√§ngen √ºber verschiedene Nullstellenmengen")
plt.yscale("log")  # Log-Skala f√ºr bessere Sichtbarkeit
plt.grid()
plt.show()
```

```python
# ‚úÖ Import der ben√∂tigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Manuelle Eingabe der verifizierten Koh√§renzl√§ngen f√ºr verschiedene N-Werte
N_values = np.array([10000, 2000000])
L_values = np.array([112.736191, 488.690558])

# ‚úÖ Numerische Ableitung dL/dN (Differenzenquotient)
dL_dN = np.diff(L_values) / np.diff(N_values)
N_midpoints = (N_values[:-1] + N_values[1:]) / 2  # Mittelpunkte f√ºr Ableitung

# ‚úÖ Modell 1: Exponentieller Zerfall ‚Üí dL/dN = -B * L
def exp_decay(N, B):
    return -B * L_values[0] * np.exp(-B * (N - N_values[0]))

# ‚úÖ Modell 2: Potenzgesetz ‚Üí dL/dN = -Œ≤ * Œ± * N^(-Œ≤ -1)
def power_law(N, beta):
    return -beta * (L_values[0] * (N / N_values[0]) ** (-beta - 1))

# ‚úÖ Modell 3: Hyperbolischer Zerfall ‚Üí dL/dN = -A / (B + N)^2
def hyperbolic_decay(N, A, B):
    return -A / (B + N) ** 2

# ‚úÖ Curve-Fit f√ºr alle drei Modelle
popt_exp, _ = curve_fit(exp_decay, N_midpoints, dL_dN, maxfev=10000)
popt_power, _ = curve_fit(power_law, N_midpoints, dL_dN, maxfev=10000)
popt_hyperbolic, _ = curve_fit(hyperbolic_decay, N_midpoints, dL_dN, maxfev=10000)

# ‚úÖ Berechnung der Fit-Werte
N_fit = np.linspace(N_values[0], N_values[1], 100)
dL_fit_exp = exp_decay(N_fit, *popt_exp)
dL_fit_power = power_law(N_fit, *popt_power)
dL_fit_hyperbolic = hyperbolic_decay(N_fit, *popt_hyperbolic)

# ‚úÖ Visualisierung der Modelle
plt.figure(figsize=(10, 6))
plt.scatter(N_midpoints, dL_dN, color='black', label="Numerische Ableitung")
plt.plot(N_fit, dL_fit_exp, 'r--', label=f"Exponentiell (B={popt_exp[0]:.6f})")
plt.plot(N_fit, dL_fit_power, 'b--', label=f"Potenzgesetz (Œ≤={popt_power[0]:.6f})")
plt.plot(N_fit, dL_fit_hyperbolic, 'g--', label=f"Hyperbolisch (A={popt_hyperbolic[0]:.6f}, B={popt_hyperbolic[1]:.6f})")

plt.xlabel("N (Anzahl Nullstellen)")
plt.ylabel("dL/dN (Ableitung der Koh√§renzl√§nge)")
plt.xscale("log")
plt.legend()
plt.grid()
plt.title("Vergleich der Modelle f√ºr dL/dN")
plt.show()

# ‚úÖ Ergebnisse ausgeben
print(f"üöÄ Exponentieller Fit: B = {popt_exp[0]:.6f}")
print(f"üöÄ Potenzgesetz Fit: Œ≤ = {popt_power[0]:.6f}")
print(f"üöÄ Hyperbolischer Fit: A = {popt_hyperbolic[0]:.6f}, B = {popt_hyperbolic[1]:.6f}")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Koh√§renzl√§ngen-Werte aus den Nullstellen-Daten
N_values = np.array([10_000, 2_000_000])
L_values = np.array([112.736191, 488.690558])

# ‚úÖ Interpolation f√ºr stabilere Ableitung
N_fine = np.logspace(np.log10(N_values[0]), np.log10(N_values[-1]), 100)
L_fine = np.interp(N_fine, N_values, L_values)

# ‚úÖ Numerische Ableitung dL/dN berechnen
dL_dN = np.gradient(L_fine, N_fine)

# ‚úÖ Transformation zur Stabilisierung
log_N_fine = np.log(N_fine)
log_dL_dN = np.log(dL_dN + 1e-10)  # Kleine Konstante, um log(0) zu vermeiden

# ‚úÖ Modelle f√ºr die Fits definieren (stabilere Formen)
def exp_decay(N, A, B):
    return A * np.exp(B * N)

def power_law(N, alpha, beta):
    return alpha * N**(-beta)

def log_power_law(log_N, log_alpha, beta):
    return log_alpha - beta * log_N  # Logarithmierte Form

def hyperbolic_decay(N, A, B):
    return A / (1 + B * N)

# ‚úÖ Curve-Fit f√ºr exponentiellen und Potenzgesetz-Fit mit stabilisierten Daten
try:
    # Exponentieller Fit bleibt unver√§ndert
    popt_exp, _ = curve_fit(exp_decay, N_fine, dL_dN, maxfev=10000)

    # Potenzgesetz-Fit auf logarithmierter Skala
    popt_power_log, _ = curve_fit(log_power_law, log_N_fine, log_dL_dN, maxfev=10000)
    alpha_stabilized = np.exp(popt_power_log[0])  # R√ºcktransformation
    beta_stabilized = popt_power_log[1]

    # Hyperbolischer Fit mit Schutz gegen Instabilit√§t
    popt_hyperbolic, _ = curve_fit(hyperbolic_decay, N_fine, dL_dN, maxfev=10000)

    print(f"üöÄ Exponentieller Fit: B = {popt_exp[1]:.6f}")
    print(f"üöÄ Potenzgesetz Fit (log-stabilisiert): Œ± = {alpha_stabilized:.6f}, Œ≤ = {beta_stabilized:.6f}")
    print(f"üöÄ Hyperbolischer Fit: A = {popt_hyperbolic[0]:.6f}, B = {popt_hyperbolic[1]:.6f}")

except Exception as e:
    print(f"‚ö†Ô∏è Fehler beim Fit: {e}")

# ‚úÖ Visualisierung der Ableitungen & Fits
plt.figure(figsize=(10,6))
plt.scatter(N_fine, dL_dN, label="Numerische Ableitung", color="black", alpha=0.5)
plt.plot(N_fine, exp_decay(N_fine, *popt_exp), label="Exp. Fit", linestyle="dashed", color="red")
plt.plot(N_fine, np.exp(log_power_law(log_N_fine, *popt_power_log)), label="Potenzgesetz Fit (log)", linestyle="dashed", color="blue")
plt.plot(N_fine, hyperbolic_decay(N_fine, *popt_hyperbolic), label="Hyperbolischer Fit", linestyle="dashed", color="green")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N (Anzahl Nullstellen)")
plt.ylabel("dL/dN (Ableitung der Koh√§renzl√§nge)")
plt.legend()
plt.title("Vergleich der Fits f√ºr die Ableitung dL/dN (stabilisiert)")
plt.show()
```

```python
# ‚úÖ Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft
import requests

# ‚úÖ Funktion zum Laden der Nullstellen
def load_zeros(url, num_rows=None):
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    response = requests.get(url)
    data = response.text.strip().split("\n")
    zeros = np.array([float(line.split()[0]) for line in data if line.strip().replace(".", "").replace("-", "").isdigit()])

    if num_rows:
        return zeros[:num_rows]
    return zeros

# ‚úÖ Fourier-Analyse zur Berechnung der Koh√§renzl√§nge
def calculate_coherence_length(zeros):
    diff = np.diff(zeros)
    norm_diff = diff / np.mean(diff)
    fourier = np.abs(fft(norm_diff))
    coherence_length = np.mean(fourier[:len(fourier)//2])
    return coherence_length

# ‚úÖ URLs der Nullstellen-Daten
url_2m = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

# ‚úÖ Lade die vollst√§ndige 2 Mio Nullstellen-Liste
zeros_full = load_zeros(url_2m)

# ‚úÖ Definiere Paketgr√∂√üen
packet_sizes = [10_000, 50_000, 100_000, 250_000, 500_000, 1_000_000, 1_500_000, 2_000_000]

# ‚úÖ Berechne Koh√§renzl√§ngen f√ºr verschiedene Paketgr√∂√üen
coherence_results = {}
for size in packet_sizes:
    zeros_subset = zeros_full[:size]  # Erstes `size`-Elemente nehmen
    coherence_length = calculate_coherence_length(zeros_subset)
    coherence_results[size] = coherence_length
    print(f"üìè Koh√§renzl√§nge ({size} Nullstellen): {coherence_length:.6f}")

# ‚úÖ Visualisierung der Koh√§renzl√§ngen
plt.figure(figsize=(8,5))
plt.plot(list(coherence_results.keys()), list(coherence_results.values()), 'o-', label="Koh√§renzl√§nge")
plt.xscale("log")
plt.xlabel("Anzahl der Nullstellen (log)")
plt.ylabel("Koh√§renzl√§nge")
plt.title("Koh√§renzl√§nge als Funktion der Nullstellen-Anzahl")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ‚úÖ Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq
import requests

# ‚úÖ Funktion zum Laden der Nullstellen von Odlyzko
def load_zeros(url, max_lines=None):
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    response = requests.get(url)
    if response.status_code == 200:
        lines = response.text.splitlines()
        zeros = []
        for line in lines:
            try:
                zeros.append(float(line.strip()))
                if max_lines and len(zeros) >= max_lines:
                    break
            except ValueError:
                continue  # √úberspringt nicht-numerische Zeilen
        print(f"‚úÖ Daten geladen: {len(zeros)} Nullstellen.")
        return np.array(zeros)
    else:
        print(f"‚ùå Fehler beim Laden der Daten.")
        return None

# ‚úÖ URLs der Nullstellen-Daten (2 Millionen Stellen)
urls = {
    "10.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "100.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "500.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "1.000.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "2.000.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

# ‚úÖ Berechnung der Koh√§renzl√§nge
def calculate_coherence_length(zeros):
    diff = np.diff(zeros)  # Abst√§nde der Nullstellen berechnen
    norm_diff = diff / np.mean(diff)  # Normalisierung
    fourier = np.abs(fft(norm_diff))  # Fourier-Transformation
    freq = fftfreq(len(norm_diff))
    coherence_length = np.mean(fourier[:len(fourier)//2])  # Mittlere Amplitude
    return coherence_length

# ‚úÖ Fit-Funktionen f√ºr Zerfallsgesetze
def power_law(N, alpha, beta):
    return alpha * N**beta

def exp_decay(N, A, B):
    return A * np.exp(B * N)

def hyperbolic_decay(N, A, B):
    return A / (N + B)

# ‚úÖ Berechnung der Koh√§renzl√§ngen f√ºr verschiedene Nullstellenmengen
coherence_results = {}
N_values = []
L_values = []

for label, url in urls.items():
    zeros = load_zeros(url, max_lines=int(label.split()[0].replace(".", "")))
    if zeros is not None:
        L = calculate_coherence_length(zeros)
        N = len(zeros)
        coherence_results[label] = L
        N_values.append(N)
        L_values.append(L)

# ‚úÖ Fit der Freese-Formel und Vergleich mit anderen Modellen
N_values = np.array(N_values)
L_values = np.array(L_values)

popt_power, _ = curve_fit(power_law, N_values, L_values)
popt_exp, _ = curve_fit(exp_decay, N_values, L_values, maxfev=10000)
popt_hyperbolic, _ = curve_fit(hyperbolic_decay, N_values, L_values)

# ‚úÖ Fit-Ergebnisse ausgeben
print("\nüöÄ **Fit-Ergebnisse:**")
print(f"‚úÖ **Potenzgesetz (Freese-Formel):** Œ± = {popt_power[0]:.6f}, Œ≤ = {popt_power[1]:.6f}")
print(f"‚úÖ **Exponentieller Zerfall:** A = {popt_exp[0]:.6f}, B = {popt_exp[1]:.6f}")
print(f"‚úÖ **Hyperbolischer Zerfall:** A = {popt_hyperbolic[0]:.6f}, B = {popt_hyperbolic[1]:.6f}")

# ‚úÖ Visualisierung der Ergebnisse
plt.figure(figsize=(8, 5))
plt.scatter(N_values, L_values, color="red", label="Daten")
plt.plot(N_values, power_law(N_values, *popt_power), label=f"Potenzgesetz (Œ≤ = {popt_power[1]:.3f})", linestyle="dashed")
plt.plot(N_values, exp_decay(N_values, *popt_exp), label="Exponentieller Zerfall", linestyle="dotted")
plt.plot(N_values, hyperbolic_decay(N_values, *popt_hyperbolic), label="Hyperbolischer Zerfall", linestyle="dashdot")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Anzahl Nullstellen N")
plt.ylabel("Koh√§renzl√§nge L(N)")
plt.legend()
plt.title("Vergleich: Potenzgesetz vs. andere Modelle")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Gegebene Werte der Nullstellen-Anzahl und Koh√§renzl√§ngen
N_values_extended = np.array([10**3, 10**4, 5 * 10**5, 10**6, 2 * 10**6])
L_values_extended = np.array([50.0, 112.7362, 250.5, 390.2, 488.6906])

# Definierte alternative Modelle

# 1Ô∏è‚É£ Potenzgesetz-Modell (Freese-Formel)
def power_law(N, alpha, beta):
    return alpha * N**beta

# 2Ô∏è‚É£ Exponentielles Modell
def exponential_model(N, A, B, C):
    return A * np.exp(-B * N) + C

# 3Ô∏è‚É£ Logarithmisches Modell
def logarithmic_model(N, a, b):
    return a + b * np.log(N)

# 4Ô∏è‚É£ Kombiniertes Modell: Potenzgesetz + log-Korrektur
def combined_model(N, alpha, beta, gamma):
    return alpha * N**beta + gamma * np.log(N)

# Optimierung der Modelle
popt_power, _ = curve_fit(power_law, N_values_extended, L_values_extended, maxfev=10000)
popt_exp, _ = curve_fit(exponential_model, N_values_extended, L_values_extended, p0=[100, 1e-6, 50], maxfev=10000)
popt_log, _ = curve_fit(logarithmic_model, N_values_extended, L_values_extended, maxfev=10000)
popt_combined, _ = curve_fit(combined_model, N_values_extended, L_values_extended, maxfev=10000)

# Berechnung der Anpassungen
N_fit = np.logspace(3, 7, 100)
L_power_fit = power_law(N_fit, *popt_power)
L_exp_fit = exponential_model(N_fit, *popt_exp)
L_log_fit = logarithmic_model(N_fit, *popt_log)
L_combined_fit = combined_model(N_fit, *popt_combined)

# Visualisierung der Fit-Kurven mit den Datenpunkten
plt.figure(figsize=(10, 7))
plt.scatter(N_values_extended, L_values_extended, color='red', label='Daten')
plt.plot(N_fit, L_power_fit, 'b--', label='Potenzgesetz')
plt.plot(N_fit, L_exp_fit, 'g-.', label='Exponentiell')
plt.plot(N_fit, L_log_fit, 'r-', label='Logarithmisch')
plt.plot(N_fit, L_combined_fit, 'm:', label='Kombiniert')

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Anzahl der Nullstellen $N$ (log)")
plt.ylabel("Koh√§renzl√§nge $L(N)$ (log)")
plt.title("Vergleich alternativer Modelle zur Beschreibung von $L(N)$")
plt.legend()
plt.grid(True)

# Anzeige der Grafik
plt.show()

# Berechnung der R¬≤-Werte zur Bewertung der Modelle
def r_squared(y_true, y_pred):
    residuals = y_true - y_pred
    ss_res = np.sum(residuals**2)
    ss_tot = np.sum((y_true - np.mean(y_true))**2)
    return 1 - (ss_res / ss_tot)

r2_power = r_squared(L_values_extended, power_law(N_values_extended, *popt_power))
r2_exp = r_squared(L_values_extended, exponential_model(N_values_extended, *popt_exp))
r2_log = r_squared(L_values_extended, logarithmic_model(N_values_extended, *popt_log))
r2_combined = r_squared(L_values_extended, combined_model(N_values_extended, *popt_combined))

# Ausgabe der Modellbewertungen
{
    "Potenzgesetz": {"Alpha": popt_power[0], "Beta": popt_power[1], "R¬≤": r2_power},
    "Exponentiell": {"A": popt_exp[0], "B": popt_exp[1], "C": popt_exp[2], "R¬≤": r2_exp},
    "Logarithmisch": {"a": popt_log[0], "b": popt_log[1], "R¬≤": r2_log},
    "Kombiniert": {"Alpha": popt_combined[0], "Beta": popt_combined[1], "Gamma": popt_combined[2], "R¬≤": r2_combined}
}
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma
from scipy.fft import fft, fftfreq

# üîπ Definition der Riemann-Siegel-Theta-Funktion
def theta_function(t):
    return np.angle(gamma(0.25 + 0.5j * t)) - 0.5 * t * np.log(np.pi)

# üîπ Lade experimentell gemessene Nullstellen-Daten (Beispiel f√ºr 10.000 Werte)
t_values = np.linspace(10, 10**6, 10000)  # Simulierte t-Werte f√ºr die Analyse

# üîπ Berechne Theta-Funktion f√ºr diese Werte
theta_values = theta_function(t_values)

# üîπ Fourier-Analyse der Theta-Oszillationen
theta_fft = np.abs(fft(theta_values))
freqs = fftfreq(len(t_values), d=(t_values[1] - t_values[0]))

# üîπ Visualisierung der Theta-Funktion und ihrer Fourier-Transformation
plt.figure(figsize=(12, 6))

# (a) Theta-Funktion
plt.subplot(1, 2, 1)
plt.plot(t_values, theta_values, label=r'$\Theta(t)$', color="blue")
plt.xlabel(r'$t$ (Nullstellen der $\zeta$-Funktion)')
plt.ylabel(r'$\Theta(t)$')
plt.title("Riemann-Siegel-Theta-Funktion")
plt.legend()
plt.grid()

# (b) Fourier-Transformation der Theta-Oszillationen
plt.subplot(1, 2, 2)
plt.plot(freqs[:len(freqs)//2], theta_fft[:len(freqs)//2], color="red")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Theta-Funktion")
plt.grid()

plt.tight_layout()
plt.show()
```
