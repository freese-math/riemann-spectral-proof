```python
# ‚úÖ Operator-Form testen: Schr√∂dinger-Hamiltonian mit FFF-Potential
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg

# ‚úÖ Parameter aus Curve-Fitting √ºbernehmen
A, B, C, w, phi = 25.95, -28.79, 0.53, 0.01, 1.63

# ‚úÖ Potenzial V(x) basierend auf der Freese-Formel
def potential(x):
    return A * x**(-1/2) + B * x**(-1) + C + w * np.cos(w * x + phi)

# ‚úÖ Diskretisierung f√ºr Schr√∂dinger-Gleichung
N = 500  # Anzahl der Zust√§nde
x_vals = np.linspace(1, N, N)  # Werte f√ºr x

# ‚úÖ Konstruiere die diskrete Schr√∂dinger-Matrix
dx = x_vals[1] - x_vals[0]
H = np.zeros((N, N))

for i in range(N):
    H[i, i] = potential(x_vals[i])  # Diagonal: Potential V(x)
    if i > 0:
        H[i, i-1] = H[i-1, i] = -1/(dx**2)  # Kinetischer Term (-d¬≤/dx¬≤)

# ‚úÖ Eigenwerte berechnen (spektrale Struktur)
eigenvalues, eigenvectors = np.linalg.eigh(H)

# ‚úÖ Ergebnisse plotten
plt.figure(figsize=(10, 5))
plt.plot(x_vals, potential(x_vals), label="Freese-Potential $V(x)$", color="blue")
plt.scatter(x_vals, eigenvalues, color="red", s=5, label="Spektrale Nullstellenstruktur")
plt.xlabel("x")
plt.ylabel("Energie / Nullstellenabstand")
plt.title("Spektrale Operator-Struktur der Freese-Formel")
plt.legend()
plt.show()

print(f"‚úÖ Erste 10 Eigenwerte des Operators: {eigenvalues[:10]}")
```

```python
# üìå Numerische L√∂sung des Operators f√ºr die Zeta-Nullstellen
import numpy as np
import matplotlib.pyplot as plt
import scipy.sparse as sp
import scipy.sparse.linalg as spla

# ‚úÖ Diskretisierung des Schr√∂dinger-Operators
def schroedinger_operator(N, dx, A=25.95, B=-28.79, C=0.53):
    """ Erstellt die Matrixdarstellung des Operators H = -d¬≤/dx¬≤ + V(x) """

    # Finite-Differenzen-Diskretisierung f√ºr -d¬≤/dx¬≤
    diag_main = np.ones(N) * (-2.0 / dx**2)
    diag_upper = np.ones(N-1) * (1.0 / dx**2)

    # Tridiagonal-Matrix f√ºr den kinetischen Term
    H_kin = sp.diags([diag_upper, diag_main, diag_upper], [-1, 0, 1])

    # Potenzialterm V(x) = A/x^(1/2) + B/x + C
    x_vals = np.linspace(1, N, N) * dx
    V_x = A * x_vals**(-1/2) + B * x_vals**(-1) + C

    # Gesamt-Hamiltonian H = -d¬≤/dx¬≤ + V(x)
    H = H_kin + sp.diags(V_x, 0)

    return H

# ‚úÖ Parameter der Berechnung
N = 500  # Anzahl der Diskretisierungspunkte
dx = 0.1  # Schrittweite f√ºr x

# ‚úÖ Erzeuge den Hamilton-Operator
H = schroedinger_operator(N, dx)

# ‚úÖ Berechnung der Eigenwerte
eigenvalues, eigenvectors = spla.eigs(H, k=20, which='SR')  # Die 20 kleinsten Eigenwerte
eigenvalues = np.real(eigenvalues)  # Nur die reellen Anteile

# üéØ Zeige die berechneten Eigenwerte
print(f"‚úÖ Erste 10 Eigenwerte des Operators: {eigenvalues[:10]}")

# ‚úÖ Plot: Potenzial & Spektrum
x_vals = np.linspace(1, N, N) * dx
V_x = 25.95 * x_vals**(-1/2) - 28.79 * x_vals**(-1) + 0.53

plt.figure(figsize=(8,5))
plt.plot(x_vals, V_x, label="Freese-Potenzial $V(x)$", color='b')
plt.scatter(np.arange(1, len(eigenvalues)+1), eigenvalues, color='r', marker='o', label="Spektrale Nullstellenstruktur")
plt.xlabel("x")
plt.ylabel("Energie / Nullstellenabstand")
plt.title("Spektrale Operator-Struktur der Freese-Formel")
plt.legend()
plt.show()

# ‚úÖ Vergleich mit Zufallsmatrix-Theorie (GUE)
plt.figure(figsize=(8,5))
plt.hist(eigenvalues, bins=30, density=True, alpha=0.6, label="Eigenwerte (FFF-Operator)")
s_vals = np.linspace(0, max(eigenvalues), 100)
gue_dist = (np.pi / 2) * s_vals * np.exp(- (np.pi / 4) * s_vals**2)
plt.plot(s_vals, gue_dist, 'r', label="Wigner-Dyson (GUE)")
plt.xlabel("Normalisierte Abst√§nde s")
plt.ylabel("Dichte")
plt.title("Vergleich der Eigenwerte mit Zufallsmatrix-Theorie")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy.optimize import minimize

# Odlyzko Nullstellen laden
zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:1000]

# Zielfunktion zur Minimierung des Fehlers zwischen Operator-Eigenwerten und Zeta-Nullstellen
def error_function(params):
    A, B, C, W, w, phi = params
    N = len(zeta_zeros)
    x = np.linspace(1, N, N)
    V = A * x**-B + B * x**-1 + C + W * np.cos(w * np.log(x) + phi)
    d = 2*np.ones(N) + V
    e = -1*np.ones(N-1)
    eigenvalues, _ = eigh_tridiagonal(d, e)
    return np.sum((eigenvalues - zeta_zeros)**2)

# Startwerte f√ºr die Optimierung
initial_params = [1.6, 0.93, 100, 0.08, np.pi, 0.5]
opt_result = minimize(error_function, initial_params, method='Nelder-Mead')

# Optimierte Parameter
A_opt, B_opt, C_opt, W_opt, w_opt, phi_opt = opt_result.x
print("Optimierte Parameter:", A_opt, B_opt, C_opt, W_opt, w_opt, phi_opt)

# Berechnung mit optimierten Werten
N = len(zeta_zeros)
x = np.linspace(1, N, N)
V_opt = A_opt * x**-B_opt + B_opt * x**-1 + C_opt + W_opt * np.cos(w_opt * np.log(x) + phi_opt)
d_opt = 2*np.ones(N) + V_opt
e_opt = -1*np.ones(N-1)
eigenvalues_opt, _ = eigh_tridiagonal(d_opt, e_opt)

# Plot der verbesserten √úbereinstimmung
plt.figure(figsize=(8,5))
plt.plot(eigenvalues_opt, 'bo', label="Operator-Eigenwerte")
plt.plot(zeta_zeros, 'ro', label="Riemann-Nullstellen")
plt.xlabel("Index n")
plt.ylabel("Eigenwerte / Nullstellen")
plt.legend()
plt.title("Optimierter Vergleich: Operator-Eigenwerte vs. Riemann-Nullstellen")
plt.show()
```
