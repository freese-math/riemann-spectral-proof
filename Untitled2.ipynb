```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def adaptive_curve_fit(func, xdata, ydata, p0, bounds=(-np.inf, np.inf), max_iterations=100000, initial_maxfev=1000, factor=2):
    maxfev = initial_maxfev
    for _ in range(10):
        try:
            params, pcov = curve_fit(func, xdata, ydata, p0=p0, bounds=bounds, maxfev=maxfev, method='trf')
            return params, pcov, maxfev
        except RuntimeError:
            print(f"Anpassung fehlgeschlagen mit maxfev={maxfev}. Erhöhe auf {maxfev*factor}.")
            maxfev *= factor
            if maxfev > max_iterations:
                print(f"Maximale Iterationsanzahl von {max_iterations} erreicht.")
                raise
    raise RuntimeError("Anpassung konnte nicht durchgeführt werden.")

# Daten
x_data = np.array([1e2, 1e4, 2e6])
y_data = np.array([6.07e19, 112.7362, 488.6906])

# Logarithmische Transformation
log_x = np.log10(x_data)
log_y = np.log10(y_data)

def log_power_law(x, a, b):
    return a - b * x

try:
    params, pcov, used_maxfev = adaptive_curve_fit(log_power_law, log_x, log_y, p0=[20, 1], bounds=([0, 0], [30, 10]))
    print(f"Erfolgreiche Anpassung mit {used_maxfev} Iterationen.")
    a, b = params
    print(f"Parameter: a={a:.3f}, b={b:.3f}")
    print(f"Potenzgesetz: y = {10**a:.3e} * x^(-{b:.3f})")

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.loglog(x_data, y_data, 'ro', label='Daten')
    x_fit = np.logspace(np.log10(x_data.min()), np.log10(x_data.max()), 100)
    y_fit = 10**(a - b * np.log10(x_fit))
    plt.loglog(x_fit, y_fit, 'b-', label='Angepasstes Potenzgesetz')
    plt.xlabel('N (Nullstellen)')
    plt.ylabel('L(N) (Kohärenzlänge)')
    plt.legend()
    plt.title('Anpassung des Potenzgesetzes an die Kohärenzlängen-Daten')
    plt.grid(True)
    plt.show()

except RuntimeError as e:
    print(f"Anpassung fehlgeschlagen: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import requests

# Funktion zum Laden der Nullstellen-Daten aus einer URL
def load_zeros(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.text.splitlines()
        zeros = [float(line.strip()) for line in data if line.strip()]
        return np.array(zeros)
    except Exception as e:
        print(f"Fehler beim Laden der Nullstellen: {e}")
        return None

# Kohärenzlängen berechnen basierend auf den Abständen der Nullstellen
def calculate_coherence_length(zeros):
    return np.diff(zeros)

# Modelle
def powerlaw_model(f, alpha, beta):
    return alpha * f**(-beta)

def exponential_model(f, a, b, c):
    return a * np.exp(-b * f) + c

def combined_model(f, alpha, beta, a, b, c):
    return alpha * f**(-beta) + a * np.exp(-b * f) + c

# Residuen berechnen
def calculate_residuals(y_true, y_pred):
    return y_true - y_pred

# URLs mit Nullstellen-Daten
url_zeros1 = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1"
url_zeros6 = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

try:
    # Lade Nullstellen-Daten
    print("Lade Nullstellen von zeros1...")
    zeros1 = load_zeros(url_zeros1)
    if zeros1 is None or len(zeros1) == 0:
        raise ValueError("Nullstellen von zeros1 konnten nicht geladen werden.")

    print("Lade Nullstellen von zeros6...")
    zeros6 = load_zeros(url_zeros6)
    if zeros6 is None or len(zeros6) == 0:
        raise ValueError("Nullstellen von zeros6 konnten nicht geladen werden.")

    # Kombiniere beide Datensätze
    all_zeros = np.concatenate([zeros1, zeros6])

    # Berechne Kohärenzlängen aus den Abständen der Nullstellen
    coherence_lengths = calculate_coherence_length(all_zeros)

    # Frequenzen als Index definieren
    frequencies = np.arange(1, len(coherence_lengths) + 1)

    # Anpassung des Potenzgesetzes
    params_pow, _ = curve_fit(
        powerlaw_model,
        frequencies,
        coherence_lengths,
        p0=[1e3, 0.5],
        bounds=(0, np.inf)
    )

    # Anpassung des exponentiellen Modells
    params_exp, _ = curve_fit(
        exponential_model,
        frequencies,
        coherence_lengths,
        p0=[1000, 0.02, 50],
        bounds=(0, np.inf)
    )

    # Anpassung des kombinierten Modells
    params_combined, _ = curve_fit(
        combined_model,
        frequencies,
        coherence_lengths,
        p0=[1e3, 0.5, 1000, 0.02, 50],
        bounds=(0, np.inf)
    )

    # Ergebnisse anzeigen
    print("\nPotenzgesetz-Modell:")
    print(f"alpha = {params_pow[0]:.3e}, beta = {params_pow[1]:.3f}")

    print("\nExponentielles Modell:")
    print(f"a = {params_exp[0]:.3e}, b = {params_exp[1]:.3f}, c = {params_exp[2]:.3f}")

    print("\nKombiniertes Modell:")
    print(f"alpha = {params_combined[0]:.3e}, beta = {params_combined[1]:.3f}, "
          f"a = {params_combined[2]:.3e}, b = {params_combined[3]:.3f}, c = {params_combined[4]:.3f}")

    # Berechnung der Residuen für jedes Modell
    residuals_pow = calculate_residuals(
        coherence_lengths,
        powerlaw_model(frequencies, *params_pow)
    )

    residuals_exp = calculate_residuals(
        coherence_lengths,
        exponential_model(frequencies, *params_exp)
    )

    residuals_combined = calculate_residuals(
        coherence_lengths,
        combined_model(frequencies, *params_combined)
    )

    # Plotten der Ergebnisse
    plt.figure(figsize=(12, 8))

    # Originaldaten plotten
    plt.loglog(frequencies, coherence_lengths, 'ro', label='Daten')

    # Potenzgesetz plotten
    f_fit = np.logspace(np.log10(frequencies.min()), np.log10(frequencies.max()), 100)

    L_fit_pow = powerlaw_model(f_fit, *params_pow)
    plt.loglog(f_fit, L_fit_pow, 'b-', label='Potenzgesetz')

    # Exponentielles Modell plotten
    L_fit_exp = exponential_model(f_fit, *params_exp)
    plt.loglog(f_fit, L_fit_exp, 'g--', label='Exponentiell')

    # Kombiniertes Modell plotten
    L_fit_combined = combined_model(f_fit, *params_combined)

    plt.loglog(f_fit, L_fit_combined, 'm-.', label='Kombiniert')

    plt.xlabel('Frequenz (f)')

    plt.ylabel('Kohärenzlänge (L)')

    plt.legend()

    plt.grid(True)

    plt.title('Anpassung von Kohärenzlängen-Daten')

    plt.show()

except Exception as e:
    print(f"Fehler: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft
import requests

# Funktion zum Laden der Nullstellen-Daten aus der URL
def load_zeros_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.text.splitlines()
        zeros = [float(line.strip()) for line in data if line.strip()]
        return np.array(zeros)
    except Exception as e:
        print(f"Fehler beim Laden der Nullstellen: {e}")
        return None

# Berechnung der Abstände zwischen benachbarten Nullstellen
def calculate_differences(zeros):
    return np.diff(zeros)

# Normalisierung der Differenzen
def normalize_differences(differences):
    return differences / np.mean(differences)

# Fourier-Analyse auf die Abstände anwenden
def perform_fourier_analysis(differences):
    N = len(differences)
    frequencies = np.fft.fftfreq(N)[:N // 2]  # Positive Frequenzen
    fft_values = np.abs(fft(differences))[:N // 2]  # Amplituden der Fourier-Transformierten
    return frequencies, fft_values

# Exponentielles Modell
def exponential_model(f, a, b, c):
    return a * np.exp(-b * f) + c

# Potenzgesetz-Modell
def powerlaw_model(f, a, b):
    return a * f**b

# URL mit den ersten 2 Millionen Nullstellen
url_zeros6 = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

# Lade die Nullstellen-Daten
print("Lade die ersten 2 Millionen Nullstellen...")
zeta_zeros = load_zeros_from_url(url_zeros6)

if zeta_zeros is not None:
    print(f"Erfolgreich {len(zeta_zeros)} Nullstellen geladen.")

    # Verwende nur einen Teil der Daten (z.B., erste 100000 Nullstellen)
    zeta_zeros_subset = zeta_zeros[:100000]

    # Berechnung der Abstände zwischen benachbarten Nullstellen
    differences = calculate_differences(zeta_zeros_subset)

    # Normalisierung der Differenzen
    normalized_differences = normalize_differences(differences)

    # Fourier-Analyse auf die normalisierten Differenzen anwenden
    frequencies, fft_values = perform_fourier_analysis(normalized_differences)

    # Filtere niedrige Frequenzen aus
    valid_indices = frequencies > 0
    frequencies = frequencies[valid_indices]
    fft_values = fft_values[valid_indices]

    # Fit-Modelle anwenden
    try:
        params_exp, _ = curve_fit(exponential_model, frequencies, fft_values, p0=[1000, 0.01, 10], bounds=(0, np.inf))
        params_pow, _ = curve_fit(powerlaw_model, frequencies, fft_values, p0=[10, -1])

        # Ergebnisse anzeigen
        print("\nExponentielles Modell:")
        print(f"L(f) = {params_exp[0]:.4f} * exp(-{params_exp[1]:.4f} * f) + {params_exp[2]:.4f}")

        print("\nPotenzgesetz-Modell:")
        print(f"L(f) = {params_pow[0]:.4f} * f^{params_pow[1]:.4f}")

        # Visualisierung der Fourier-Analyse und Fit-Modelle
        plt.figure(figsize=(12, 8))

        # Original FFT-Daten plotten
        plt.plot(frequencies, fft_values, 'ro', label='Fourier-Amplituden')

        # Exponentielles Modell plotten
        plt.plot(frequencies, exponential_model(frequencies, *params_exp), 'g--', label='Exponentielles Modell')

        # Potenzgesetz-Modell plotten
        plt.plot(frequencies, powerlaw_model(frequencies, *params_pow), 'b-', label='Potenzgesetz-Modell')

        plt.xlabel('Frequenz (f)')
        plt.ylabel('Amplitude')
        plt.title('Fourier-Analyse und Fit-Modelle')
        plt.legend()
        plt.grid(True)

        plt.show()

    except Exception as e:
        print(f"Fehler bei der Anpassung der Modelle: {e}")

else:
    print("Fehler: Die Nullstellen konnten nicht geladen werden.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score
import requests

# Funktion zum Laden der Nullstellen-Daten aus der URL
def load_zeros_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.text.splitlines()
        zeros = [float(line.strip()) for line in data if line.strip()]
        return np.array(zeros)
    except Exception as e:
        print(f"Fehler beim Laden der Nullstellen: {e}")
        return None

# Berechnung der Abstände zwischen benachbarten Nullstellen
def calculate_differences(zeros):
    return np.diff(zeros)

# Fourier-basierte Frequenzen simulieren (falls nötig)
def generate_frequencies(differences):
    N = len(differences)
    frequencies = np.linspace(1, N, N)
    return frequencies

# Exponentielles Modell
def exponential_model(f, a, b, c):
    return a * np.exp(-b * f) + c

# Potenzgesetz-Modell
def powerlaw_model(f, a, b):
    return a * f**b

# Logarithmisches Modell
def logarithmic_model(f, a, b):
    return a * np.log(f) + b

# URL mit den ersten 2 Millionen Nullstellen
url_zeros6 = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

# Lade die Nullstellen-Daten
print("Lade die ersten 2 Millionen Nullstellen...")
zeta_zeros = load_zeros_from_url(url_zeros6)

if zeta_zeros is not None:
    print(f"Erfolgreich {len(zeta_zeros)} Nullstellen geladen.")

    # Berechnung der Abstände zwischen benachbarten Nullstellen
    differences = calculate_differences(zeta_zeros)

    # Frequenzen generieren (simuliert)
    frequencies = generate_frequencies(differences)

    # Fit-Modelle anwenden
    try:
        # Exponentielles Modell
        params_exp, _ = curve_fit(exponential_model, frequencies, differences, p0=[1000, 0.01, 50])
        exp_fit = exponential_model(frequencies, *params_exp)
        r2_exp = r2_score(differences, exp_fit)

        # Potenzgesetz-Modell
        params_pow, _ = curve_fit(powerlaw_model, frequencies, differences)
        pow_fit = powerlaw_model(frequencies, *params_pow)
        r2_pow = r2_score(differences, pow_fit)

        # Logarithmisches Modell
        params_log, _ = curve_fit(logarithmic_model, frequencies[frequencies > 0], differences[frequencies > 0])
        log_fit = logarithmic_model(frequencies[frequencies > 0], *params_log)
        r2_log = r2_score(differences[frequencies > 0], log_fit)

        # Ergebnisse anzeigen
        print("\nExponentielles Modell:")
        print(f"L(f) = {params_exp[0]:.4f} * exp(-{params_exp[1]:.4f} * f) + {params_exp[2]:.4f}")
        print(f"R²-Wert: {r2_exp:.4f}")

        print("\nPotenzgesetz-Modell:")
        print(f"L(f) = {params_pow[0]:.4f} * f^{params_pow[1]:.4f}")
        print(f"R²-Wert: {r2_pow:.4f}")

        print("\nLogarithmisches Modell:")
        print(f"L(f) = {params_log[0]:.4f} * log(f) + {params_log[1]:.4f}")
        print(f"R²-Wert: {r2_log:.4f}")

        # Visualisierung der Modelle
        plt.figure(figsize=(12, 8))

        # Originaldaten plotten
        plt.plot(frequencies, differences, 'ro', label='Daten')

        # Exponentielles Modell plotten
        plt.plot(frequencies, exp_fit, 'g--', label='Exponentielles Modell')

        # Potenzgesetz-Modell plotten
        plt.plot(frequencies, pow_fit, 'b-', label='Potenzgesetz-Modell')

        # Logarithmisches Modell plotten
        plt.plot(frequencies[frequencies > 0], log_fit, 'm-.', label='Logarithmisches Modell')

        plt.xlabel('Frequenz (f)')
        plt.ylabel('Kohärenzlänge (L)')
        plt.title('Modellanpassung für Kohärenzlängen')
        plt.legend()
        plt.grid(True)

        plt.show()

    except Exception as e:
        print(f"Fehler bei der Anpassung der Modelle: {e}")

else:
    print("Fehler: Die Nullstellen konnten nicht geladen werden.")
```

```python
# ====================================================
# 📌 Code für Google Colab: Nullstellen der Zeta-Funktion analysieren
# ====================================================
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import StringIO
from scipy.optimize import curve_fit

# ----------------------------------------------------
# 🔹 1. Nullstellen der Zeta-Funktion von Odlyzkos Website laden
# ----------------------------------------------------
def load_riemann_zeros(N=500):
    """Lädt die ersten N Nullstellen der Riemannschen Zeta-Funktion von Odlyzkos Datenbank."""
    url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

    print(f"📥 Lade {N} Nullstellen von {url} ...")
    response = requests.get(url)

    if response.status_code != 200:
        raise ValueError("⚠️ Fehler beim Laden der Nullstellen-Daten.")

    # Textdaten einlesen und filtern
    text_data = response.text
    lines = text_data.split("\n")

    t_values = []
    for line in lines:
        try:
            value = float(line.strip())
            t_values.append(value)
            if len(t_values) >= N:
                break
        except ValueError:
            continue  # Überspringt leere oder fehlerhafte Zeilen

    if len(t_values) < N:
        raise ValueError(f"⚠️ Nur {len(t_values)} Nullstellen gefunden. Überprüfe die Quelle.")

    print(f"✅ Erfolgreich {len(t_values)} Nullstellen geladen.")
    return np.array(t_values)

# ----------------------------------------------------
# 🔹 2. Berechnung der spektralen Kohärenzlänge L(f)
# ----------------------------------------------------
def compute_coherence_length(t_values, f_range):
    """Berechnet die spektrale Kohärenzlänge L(f) für gegebene Nullstellen."""
    L_values = []
    for f in f_range:
        L_f = np.abs(np.sum(np.exp(-1j * t_values * f)))
        L_values.append(L_f)
    return np.array(L_values)

# ----------------------------------------------------
# 🔹 3. Power-Law Fit: L(f) = alpha * f^(-beta)
# ----------------------------------------------------
def power_law(f, alpha, beta):
    return alpha * f**(-beta)

# ----------------------------------------------------
# 🔹 4. Hauptprogramm: Nullstellen laden, Kohärenzlänge berechnen & Plotten
# ----------------------------------------------------
def main():
    N = 500  # Anzahl der Nullstellen, die analysiert werden sollen
    t_values = load_riemann_zeros(N)  # Nullstellen laden

    # Frequenzbereich logarithmisch definieren
    f_range = np.logspace(0, 4, 100)

    # Kohärenzlänge berechnen
    L_values = compute_coherence_length(t_values, f_range)

    # Fit der Daten mit einem Potenzgesetz
    try:
        popt, pcov = curve_fit(power_law, f_range, L_values, p0=[100, 0.5])
        alpha_fit, beta_fit = popt
        fit_successful = True
    except RuntimeError:
        print("⚠️ Fit fehlgeschlagen. Überprüfe die Daten.")
        fit_successful = False
        alpha_fit, beta_fit = np.nan, np.nan

    # Plot der Ergebnisse
    plt.figure(figsize=(8,6))
    plt.loglog(f_range, L_values, 'ro', label="Daten")
    if fit_successful:
        plt.loglog(f_range, power_law(f_range, *popt), 'b--',
                   label=f"Fit: α={alpha_fit:.2f}, β={beta_fit:.2f}")
    plt.xlabel("Frequenz (f)")
    plt.ylabel("Kohärenzlänge (L)")
    plt.legend()
    plt.title("Spektrale Kohärenzlänge der Riemann-Nullstellen")
    plt.grid()
    plt.show()

    # Ergebnisse ausgeben
    if fit_successful:
        print(f"📊 Gefittete Parameter: α = {alpha_fit:.2f}, β = {beta_fit:.2f}")
    else:
        print("⚠️ Kein gültiger Fit möglich.")

# ----------------------------------------------------
# 🔹 5. Skript ausführen
# ----------------------------------------------------
if __name__ == "__main__":
    main()
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Daten aus vorherigen Fits
N_values = np.array([500, 1000, 5000, 10000])
beta_values = np.array([0.716, 0.745, 0.800, 0.819])

# --- 1. Alternative Fit-Modelle ---
def beta_exp(N, beta_inf, c, p):
    """Exponentieller Zerfall β(N) = β∞ + c * exp(-p * N)"""
    return beta_inf + c * np.exp(-p * N)

def beta_log(N, beta_inf, c):
    """Logarithmisches Modell β(N) = β∞ - c * log(N)"""
    return beta_inf - c * np.log(N)

def beta_hyperbolic(N, beta_inf, c):
    """Hyperbolischer Zerfall β(N) = β∞ - c / N"""
    return beta_inf - c / N

# --- 2. Fits durchführen ---
fits = {
    "Exponentieller Zerfall": beta_exp,
    "Logarithmisches Modell": beta_log,
    "Hyperbolischer Zerfall": beta_hyperbolic
}

fit_results = {}
for name, model in fits.items():
    try:
        popt, _ = opt.curve_fit(model, N_values, beta_values, maxfev=10000)
        fit_results[name] = popt
        print(f"✅ {name} Fit erfolgreich: {popt}")
    except RuntimeError:
        print(f"⚠️ {name} Fit fehlgeschlagen.")

# --- 3. Visualisierung ---
plt.figure(figsize=(8,6))
plt.scatter(N_values, beta_values, label="Gemessene Werte", color="red")

# Fit-Kurven plotten
N_plot = np.linspace(500, 20000, 100)
for name, model in fits.items():
    if name in fit_results:
        plt.plot(N_plot, model(N_plot, *fit_results[name]), label=name)

plt.xscale("log")
plt.xlabel("N (Anzahl Nullstellen)")
plt.ylabel("β(N)")
plt.title("Vergleich verschiedener Fit-Modelle für β(N)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# ✅ Nullstellen-Daten laden
real_zeros_url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

def load_real_zeros(url):
    try:
        zeros = np.loadtxt(url, dtype=np.float64)  # Erzwinge 64-bit Genauigkeit
        return zeros
    except Exception as e:
        print(f"⚠ Fehler beim Laden der echten Nullstellen: {e}")
        return None

# ✅ Funktion zur Berechnung der Kohärenzlänge mit echten Differenzen
def calculate_coherence_length(zeros):
    diff = np.diff(zeros)  # Echte Differenzen berechnen
    norm_diff = diff / np.mean(diff)  # Normalisieren

    fourier = fft(norm_diff)  # Fourier-Analyse
    freq = fftfreq(len(norm_diff))  # Frequenzen berechnen

    coherence_length = np.mean(np.abs(fourier[:len(fourier)//2]))  # Mittlere Amplitude
    print(f"\n🚀 Neue Kohärenzlänge (mit echten Differenzen): {coherence_length:.6f}")

    # Visualisierung des Fourier-Spektrums
    plt.figure(figsize=(8, 4))
    plt.plot(freq[:len(freq)//2], np.abs(fourier[:len(fourier)//2]))
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.title("Fourier-Spektrum der normalisierten Differenzen")
    plt.grid()
    plt.show()

    return coherence_length

# ✅ Echte Nullstellen laden
real_zeros = load_real_zeros(real_zeros_url)
if real_zeros is None:
    raise ValueError("❌ Konnte echte Nullstellen nicht laden.")

# ✅ Berechnung der neuen Kohärenzlänge
new_coherence_length = calculate_coherence_length(real_zeros)
```

```python
import numpy as np
import requests
import scipy.fftpack
import matplotlib.pyplot as plt

# 📌 URLs der Nullstellen-Daten
odlyzko_urls = [
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros3",
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros4",
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros5",
    "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
]

# 📌 Funktion zum Laden der Nullstellen-Daten
def load_zeros(url):
    print(f"📥 Lade Daten von: {url}")
    response = requests.get(url)
    if response.status_code == 200:
        return np.array([float(line.strip()) for line in response.text.splitlines()])
    else:
        print(f"❌ Fehler beim Abrufen der Daten: {url}")
        return None

# 📌 Funktion zur Fourier-Analyse
def compute_fft(zeros, dataset_name):
    print(f"📊 Fourier-Analyse für {dataset_name}...")
    diffs = np.diff(zeros)

    # Frequenzberechnung
    freq = scipy.fftpack.fftfreq(len(diffs), d=np.median(diffs))
    spectrum = np.abs(scipy.fftpack.fft(diffs))

    # Dominante Frequenz finden
    dominant_freq = freq[np.argmax(spectrum)]
    print(f"🔍 Dominante Frequenz ({dataset_name}): {dominant_freq:.6f}")

    # 📊 Plot erstellen
    plt.figure(figsize=(10, 4))
    plt.plot(freq[:len(freq)//2], spectrum[:len(spectrum)//2])
    plt.xlabel("Frequenz")
    plt.ylabel("Spektrale Amplitude")
    plt.title(f"Fourier-Analyse der Nullstellen-Abstände ({dataset_name})")
    plt.show()

    return dominant_freq

# 📌 Hauptprogramm: Lade Daten und führe Fourier-Analyse durch
dominant_frequencies = []
for i, url in enumerate(odlyzko_urls):
    zeros = load_zeros(url)
    if zeros is not None:
        dominant_freq = compute_fft(zeros, f"zeros{i+1}")
        dominant_frequencies.append(dominant_freq)

# 📊 Ergebnisse zusammenfassen
print("\n📊 **Zusammenfassung der dominanten Frequenzen:**")
for i, freq in enumerate(dominant_frequencies):
    print(f"Dataset zeros{i+1}: {freq:.6f}")
```
