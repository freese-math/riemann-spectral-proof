```python
import numpy as np
import matplotlib.pyplot as plt
import requests

from scipy.optimize import curve_fit

# üìÇ **Datenquelle: Odlyzko Nullstellen (2 Mio)**
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"

# üìå **1Ô∏è‚É£ Lade Nullstellen aus Online-Quelle**
def lade_nullstellen(url):
    response = requests.get(url)
    if response.status_code == 200:
        nullstellen = np.loadtxt(response.iter_lines(), dtype=np.float64)
        print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")
        return nullstellen
    else:
        print("‚ùå Fehler beim Laden der Nullstellen!")
        return None

# üìå **2Ô∏è‚É£ Berechne Nullstellen-Abst√§nde**
def berechne_abstaende(nullstellen):
    return np.diff(nullstellen).astype(np.float64)  # Float-Format explizit setzen

# üìå **3Ô∏è‚É£ Freese-Formel Standard (FFS)**
def freese_standard(n, A, B, C):
    return A * np.power(n, -0.5) + B * np.power(n, -1) + C  # Float-Exponenten sichern

# üìå **4Ô∏è‚É£ Freese-Formel Oszillierend (FFO)**
def freese_oszillierend(n, A, B, C, w, phi):
    return A * np.power(n, -0.5) + B * np.power(n, -1) + C + w * np.cos(n * w + phi)

# üìå **5Ô∏è‚É£ Optimierungsfunktion f√ºr FFS**
def optimiere_freese(nullstellen_abstaende, oszillierend=False):
    n_values = np.arange(1, len(nullstellen_abstaende) + 1, dtype=np.float64)
    if oszillierend:
        p0 = [25.96, -28.80, 0.53, 0.01, 1.64]  # Startwerte f√ºr FFO
        popt, _ = curve_fit(freese_oszillierend, n_values, nullstellen_abstaende, p0=p0)
    else:
        p0 = [25.95, -28.79, 0.53]  # Startwerte f√ºr FFS
        popt, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=p0)
    return popt

# üìå **6Ô∏è‚É£ Fehlerberechnung**
def berechne_fehler(nullstellen_abstaende, modellwerte):
    fehler = np.abs(nullstellen_abstaende - modellwerte) / nullstellen_abstaende * 100
    return fehler.mean(), np.median(fehler), fehler.max()

# üìå **7Ô∏è‚É£ Visualisierung der Ergebnisse**
def plot_freese(nullstellen_abstaende, params_ffs, params_ffo):
    n_values = np.arange(1, len(nullstellen_abstaende) + 1, dtype=np.float64)
    ffs_values = freese_standard(n_values, *params_ffs)
    ffo_values = freese_oszillierend(n_values, *params_ffo)

    plt.figure(figsize=(10,5))
    plt.scatter(n_values, nullstellen_abstaende, color="black", s=1, label="Echte Nullstellen-Abst√§nde")
    plt.plot(n_values, ffs_values, "b--", label="FFS (Standard)")
    plt.plot(n_values, ffo_values, "r-", label="FFO (Oszillierend)")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Index der Nullstelle (n)")
    plt.ylabel("Abstand Œîn")
    plt.legend()
    plt.title("üìà Vergleich: Echte Nullstellen vs. FFS & FFO")
    plt.grid(True, which="both", linestyle="--")
    plt.show()

# üìå **üöÄ Hauptprogramm**
nullstellen = lade_nullstellen(URL)

if nullstellen is not None:
    nullstellen_abstaende = berechne_abstaende(nullstellen)

    # üéØ Optimiere FFS & FFO
    params_ffs = optimiere_freese(nullstellen_abstaende, oszillierend=False)
    params_ffo = optimiere_freese(nullstellen_abstaende, oszillierend=True)

    print(f"‚úÖ Optimierte FFS-Parameter: A={params_ffs[0]:.6f}, B={params_ffs[1]:.6f}, C={params_ffs[2]:.6f}")
    print(f"üìå Optimierte FFO-Parameter: A={params_ffo[0]:.6f}, B={params_ffo[1]:.6f}, C={params_ffo[2]:.6f}, w={params_ffo[3]:.6f}, œÜ={params_ffo[4]:.6f}")

    # üìä Fehlerberechnung
    n_values = np.arange(1, len(nullstellen_abstaende) + 1, dtype=np.float64)
    ffs_values = freese_standard(n_values, *params_ffs)
    ffo_values = freese_oszillierend(n_values, *params_ffo)

    fehler_ffs = berechne_fehler(nullstellen_abstaende, ffs_values)
    fehler_ffo = berechne_fehler(nullstellen_abstaende, ffo_values)

    # üìä **Fehlertabelle**
    print("\nüìä **Fehlertabelle:**")
    print(f"Modell             √ò Fehler (%)   Median Fehler (%)   Max Fehler (%)")
    print(f"FFS (Standard)     {fehler_ffs[0]:.6f}   {fehler_ffs[1]:.6f}   {fehler_ffs[2]:.6f}")
    print(f"FFO (Oszillierend) {fehler_ffo[0]:.6f}   {fehler_ffo[1]:.6f}   {fehler_ffo[2]:.6f}")

    # üî• **Plot der Ergebnisse**
    plot_freese(nullstellen_abstaende, params_ffs, params_ffo)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üìå Lade Nullstellen-Daten
def lade_nullstellen(datei="zeros6.txt"):
    try:
        nullstellen = np.loadtxt(datei)
        return nullstellen
    except Exception as e:
        print(f"Fehler beim Laden der Nullstellen: {e}")
        return None

# üéØ Berechne Nullstellen-Abst√§nde
def berechne_abstaende(nullstellen):
    return np.diff(nullstellen).astype(np.float64)  # Float explizit setzen

# üìå Fibonacci-basierte Frequenzkomponente
def fibonacci_frequenz(n, base_freq=0.618):
    return base_freq * (1 + np.cos(2 * np.pi * np.log(n) / np.log(1.618)))

# üìå Lorentz-Transformation zur Stabilisierung
def lorentz_korrektur(n, v=0.1):
    return 1 / np.sqrt(1 - (v**2 / (1 + n)))

# üìå Erweiterte Freese-Formel (FFF)
def freese_erweitert(n, A, B, C, D, E, v, f_base):
    freq_term = fibonacci_frequenz(n, f_base)
    lorentz_term = lorentz_korrektur(n, v)
    return (A * np.power(n, B) + C * np.log(n) + D / n + E * freq_term) * lorentz_term

# üéØ Optimierung der erweiterten FFF
def optimiere_freese(nullstellen_abstaende):
    n = np.arange(1, len(nullstellen_abstaende) + 1)

    # Startwerte basierend auf fr√ºheren Erkenntnissen
    p0 = [14.1347, 0.484906, 0, 0, 0, 0.1, 0.618]  # A, B, C, D, E, v, f_base

    try:
        params, _ = curve_fit(freese_erweitert, n, nullstellen_abstaende, p0=p0, maxfev=5000)
        return params
    except RuntimeError as e:
        print(f"Optimierungsfehler: {e}")
        return None

# üìä Visualisierung der optimierten FFF
def plot_freese(nullstellen_abstaende, params):
    n = np.arange(1, len(nullstellen_abstaende) + 1)
    fff_values = freese_erweitert(n, *params)

    plt.figure(figsize=(10, 6))
    plt.scatter(n, nullstellen_abstaende, s=1, color="black", alpha=0.5, label="Echte Nullstellen-Abst√§nde")
    plt.plot(n, fff_values, "r--", label="Optimierte FFF (Erweitert)")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("Index der Nullstelle (n)")
    plt.ylabel("Abstand Œîn")
    plt.title("üìä Vergleich: Echte Nullstellen vs. Optimierte FFF")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

# üî• Hauptprogramm
nullstellen = lade_nullstellen()
if nullstellen is not None:
    nullstellen_abstaende = berechne_abstaende(nullstellen)
    params_fff = optimiere_freese(nullstellen_abstaende)

    if params_fff is not None:
        print(f"‚úÖ Optimierte FFF-Parameter: A={params_fff[0]:.6f}, B={params_fff[1]:.6f}, "
              f"C={params_fff[2]:.6f}, D={params_fff[3]:.6f}, E={params_fff[4]:.6f}, "
              f"v={params_fff[5]:.6f}, f_base={params_fff[6]:.6f}")
        plot_freese(nullstellen_abstaende, params_fff)
    else:
        print("‚ö† Optimierung fehlgeschlagen.")
else:
    print("‚ö† Nullstellen-Daten konnten nicht geladen werden.")
```
