```python
# Google Colab: Berechnung der Kohärenzlänge in GUE-Zufallsmatrizen (bis N = 2000)

import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from multiprocessing import Pool, cpu_count

# Optimierte Funktion zur Erzeugung einer GUE-Zufallsmatrix
def generate_GUE_matrix(N):
    """Erzeugt eine zufällige hermitesche GUE-Matrix der Größe N x N."""
    A = np.random.randn(N, N) / np.sqrt(2) + 1j * np.random.randn(N, N) / np.sqrt(2)
    return (A + A.conj().T) / 2  # Hermitesche Matrix

# Parallele Berechnung der Kohärenzlänge
def compute_coherence_length_parallel(N, num_matrices=3):
    """Parallele Berechnung der mittleren Kohärenzlänge mit weniger Zufallsmatrizen."""
    with Pool(processes=min(4, cpu_count())) as pool:  # Nutze max. 4 Kerne, um Speicher zu schonen
        results = pool.map(generate_GUE_matrix, [N] * num_matrices)

    coherence_lengths = []

    for H in results:
        eigenvalues = np.sort(eigh(H, eigvals_only=True))  # Eigenwerte berechnen
        spacing = np.diff(eigenvalues)  # Abstand zwischen Eigenwerten
        norm_spacing = spacing / np.mean(spacing)  # Normalisierung

        coherence_length = np.mean(norm_spacing)  # Mittlere spektrale Kohärenzlänge
        coherence_lengths.append(coherence_length)

    return np.mean(coherence_lengths), np.std(coherence_lengths)

# Erweiterte Matrixgrößen für Berechnung (bis N=2000)
N_values_parallel = [10, 50, 100, 500, 1000, 1500, 2000]
coherence_results_parallel = [compute_coherence_length_parallel(N) for N in N_values_parallel]

# Ergebnisse extrahieren
L_values_parallel = [result[0] for result in coherence_results_parallel]
L_errors_parallel = [result[1] for result in coherence_results_parallel]

# Visualisierung der Ergebnisse
plt.figure(figsize=(8, 6))
plt.errorbar(N_values_parallel, L_values_parallel, yerr=L_errors_parallel, fmt='o', label='GUE-Kohärenzlänge (bis N=2000)', capsize=5)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Matrixgröße N (log)")
plt.ylabel("Mittlere Kohärenzlänge L(N) (log)")
plt.title("Parallele Berechnung der Kohärenzlänge in Zufallsmatrizen (GUE)")
plt.legend()
plt.grid(True)
plt.show()

# Ergebnisse als Tabelle ausgeben
import pandas as pd
df_gue_parallel = pd.DataFrame({"N (Matrixgröße)": N_values_parallel, "Mittlere L(N)": L_values_parallel, "Standardabweichung": L_errors_parallel})
df_gue_parallel
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Beispielwerte für N und L(N) (Echte Daten müssen hier eingefügt werden)
N = np.array([1, 2, 3, 4, 5])  # Beispiel N-Werte
L_N = np.array([0.1, 0.3, 0.5, 0.7, 1.0])  # Beispiel Kohärenzlängen L(N)

# Logarithmische Transformation
log_N = np.log(N)
log_L_N = np.log(L_N)

# Fit der logarithmischen Skala (Anpassung von log(A) und beta)
def fit_model(x, log_A, beta):
    return log_A + beta * x

# Fitting der logarithmischen Daten
params, covariance = curve_fit(fit_model, log_N, log_L_N)
log_A, beta = params

# Berechnung der Modellvorhersagen
log_L_N_fit = fit_model(log_N, log_A, beta)

# Umkehrung der Logarithmen, um den Modellwert L(N) zu erhalten
L_N_fit = np.exp(log_L_N_fit)

# Plot der Originaldaten und der Modellvorhersage
plt.figure(figsize=(8,6))
plt.plot(N, L_N, 'o', label="Originaldaten")
plt.plot(N, L_N_fit, '-', label="Modellvorhersage")
plt.xlabel('Anzahl der Nullstellen (N)')
plt.ylabel('Kohärenzlängen (L(N))')
plt.legend()
plt.title('Vergleich von Originaldaten und Modellvorhersage')
plt.show()

# Ausgabe der berechneten Parameter
print(f"Skalierungsfaktor A: {np.exp(log_A)}")
print(f"Exponent beta: {beta}")
```

```python
# 🚀 Importiere benötigte Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ✅ Funktion zur Generierung eines analytischen Operators
def generate_diagonal_operator(N, beta=0.2825):
    """
    Erstellt einen analytischen Operator mit einer Eigenwertstruktur,
    sodass die spektralen Abstände L(N) ~ N^beta skalieren.
    """
    eigenvalues_diag = np.linspace(1, N**beta, N)  # Eigenwerte basierend auf Skalierung
    spacing_diag = np.diff(eigenvalues_diag)  # Eigenwert-Abstände
    norm_spacing_diag = spacing_diag / np.mean(spacing_diag)  # Normalisierung
    coherence_length_diag = np.mean(norm_spacing_diag)  # Kohärenzlänge berechnen
    return coherence_length_diag

# ✅ Test für verschiedene Werte von N
N_values_diag = [10, 50, 100, 500, 1000, 2000, 5000, 10000]
coherence_results_diag = [generate_diagonal_operator(N) for N in N_values_diag]

# ✅ Visualisierung der Kohärenzlängen
plt.figure(figsize=(8, 6))
plt.plot(N_values_diag, coherence_results_diag, 'o-', label='Diagonaler Operator (Analytisch)')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Matrixgröße N (log)")
plt.ylabel("Mittlere Kohärenzlänge L(N) (log)")
plt.title("Spektrale Kohärenzlänge mit analytischer Eigenwertstruktur (β = 0.2825)")
plt.legend()
plt.grid(True)
plt.show()

# ✅ Ergebnisse als Tabelle ausgeben
df_diagonal_operator = pd.DataFrame({
    "N (Matrixgröße)": N_values_diag,
    "Mittlere L(N)": coherence_results_diag
})

# ✅ Zeige Tabelle in Google Colab
from google.colab import data_table
data_table.DataTable(df_diagonal_operator)

# ✅ Ausgabe der berechneten Werte für Kontrolle
print("🔍 Berechnete Kohärenzlängen:")
print(df_diagonal_operator)
```

```python
# 📌 Importiere die benötigten Pakete
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy.fftpack import fft
from scipy.stats import gaussian_kde

# 📌 Nullstellen-Datenquellen (Odlyzko)
urls = {
    "100": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "10k": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2M": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

# 📌 Funktion zum Laden der Nullstellen
def load_zeros(url):
    return np.loadtxt(url)

# 📌 Berechnung der normierten Abstände
def compute_normalized_diffs(zeros):
    diffs = np.diff(zeros)
    return diffs / np.mean(diffs)

# 📌 Berechnung der Fourier-Analyse
def compute_fourier_spectrum(norm_diffs):
    fourier = fft(norm_diffs)
    freq = np.fft.fftfreq(len(norm_diffs))
    return freq, np.abs(fourier)

# 📌 Curve-Fitting Funktion: Potenzgesetz-Modell (Freese-Formel)
def freese_formula(N, alpha, beta):
    return alpha * N ** beta

# 📌 Daten laden & berechnen
data = {}
for label, url in urls.items():
    print(f"📥 Lade Nullstellen: {label} Nullstellen...")
    data[label] = {}
    data[label]["zeros"] = load_zeros(url)
    data[label]["norm_diffs"] = compute_normalized_diffs(data[label]["zeros"])
    data[label]["freq"], data[label]["fourier"] = compute_fourier_spectrum(data[label]["norm_diffs"])

# 📌 Kohärenzlängen berechnen (Mittelwert des Fourier-Spektrums)
L_values = {}
for label, dataset in data.items():
    L_values[label] = np.mean(dataset["fourier"])

# 📌 Fit der Freese-Formel (Potenzgesetz)
N_values = np.array([10**2, 10**4, 2 * 10**6])
L_values_arr = np.array([L_values["100"], L_values["10k"], L_values["2M"]])

# 🚀 **Logarithmische Transformation für stabileres Fitten**
log_N = np.log(N_values)
log_L = np.log(L_values_arr)

# 🚀 **Neue Startwerte für die Optimierung**
initial_guess = [np.exp(np.mean(log_L)), 0.3]

# 🚀 **Curve-Fitting mit Stabilisierung**
try:
    popt, _ = curve_fit(lambda N, a, b: a * N ** b, N_values, L_values_arr, p0=initial_guess)
    alpha_opt, beta_opt = popt
except RuntimeError as e:
    print(f"⚠️ Curve-Fitting fehlgeschlagen: {e}")
    alpha_opt, beta_opt = np.nan, np.nan  # Setze NaN, falls Fit fehlschlägt

# 📌 📊 Visualisierung der Ergebnisse
plt.figure(figsize=(10, 6))
plt.scatter(N_values, L_values_arr, color="red", label="Messpunkte")
N_fit = np.logspace(2, 7, 100)
plt.plot(N_fit, freese_formula(N_fit, *popt), linestyle="--", color="blue",
         label=f"Fit: $L(N) = {alpha_opt:.3f} \cdot N^{{{beta_opt:.4f}}}$")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Anzahl der Nullstellen $N$ (log)")
plt.ylabel("Kohärenzlänge $L(N)$ (log)")
plt.title("Freese-Formel: Kohärenzlängen-Modell $L(N) = \\alpha N^{\\beta}$")
plt.legend()
plt.grid(True)

# 📌 📂 Speichern des Plots
plt.savefig("freese_kohärenzlängen_plot.png")
plt.show()

# 📌 📜 Ausgabe der Fit-Parameter
print(f"🚀 Optimierter Fit: L(N) = {alpha_opt:.3f} * N^{beta_opt:.4f}")
```
