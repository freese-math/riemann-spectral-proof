```python
# üìå Wichtige Bibliotheken importieren
import numpy as np
import scipy.fftpack
import sqlite3
import struct
import os
import matplotlib.pyplot as plt

# üìå Pfade zu den Datenbank- und Nullstellen-Dateien in Google Drive
index_db_path = "/content/drive/MyDrive/Aus Chrome gespeichert/index.db"
zeros_file_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros_30607946000.dat"

# üìå Pr√ºfen, ob Dateien vorhanden sind
if not os.path.exists(index_db_path):
    raise FileNotFoundError(f"‚ùå Fehler: Index-Datei nicht gefunden unter {index_db_path}")
if not os.path.exists(zeros_file_path):
    raise FileNotFoundError(f"‚ùå Fehler: Nullstellen-Datei nicht gefunden unter {zeros_file_path}")

print("‚úÖ Dateien gefunden! Starte Analyse...")

# üìå Funktion zur Extraktion der Nullstellen aus der SQLite-Datenbank
def get_zero_filenames_from_db(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT filename FROM zero_index")
    filenames = [row[0] for row in cursor.fetchall()]
    conn.close()
    return filenames

# üìå Funktion zum Entpacken der Nullstellen aus der .dat Datei
def read_platt_zeros(file_path):
    zeros = []

    # üìå Pr√ºfen der Dateigr√∂√üe
    file_size = os.path.getsize(file_path)
    print(f"üì¶ Nullstellen-Dateigr√∂√üe: {file_size} Bytes")

    # Muss ein Vielfaches von 8 Bytes sein (double-precision floating point)
    if file_size % 8 != 0:
        raise ValueError(f"‚ùå Fehler: Datei hat eine ung√ºltige Gr√∂√üe ({file_size} Bytes). Erwartet: Vielfaches von 8.")

    with open(file_path, "rb") as file:
        while True:
            data = file.read(8)
            if not data:
                break
            try:
                zeros.append(struct.unpack("d", data)[0])  # Bin√§rdaten in Float umwandeln
            except struct.error:
                print("‚ùå Fehler: Problem beim Entpacken der Nullstellen-Daten.")
                break

    return np.array(zeros)

# üìå 1Ô∏è‚É£ Nullstellen aus der Datei extrahieren
print("üì• Lade echte Nullstellen-Daten...")
t_zeros = read_platt_zeros(zeros_file_path)
print(f"‚úÖ {len(t_zeros)} Nullstellen erfolgreich geladen!")

# üìå 2Ô∏è‚É£ Fourier-Transformation der echten Nullstellen
print("üìä F√ºhre Fourier-Analyse durch...")
frequencies = scipy.fftpack.fftfreq(len(t_zeros), d=(t_zeros[1] - t_zeros[0]))
spectrum = np.abs(scipy.fftpack.fft(t_zeros))

# üìå 3Ô∏è‚É£ Visualisierung der Fourier-Analyse
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:len(frequencies)//2], np.log10(spectrum[:len(spectrum)//2]), color='blue', label="Fourier-Spektrum der echten Nullstellen")
plt.axvline(x=0.5, color='red', linestyle="dashed", label="RH-Kritische Frequenz $f = 1/2$")
plt.xlabel("Frequenz", fontsize=12)
plt.ylabel("Log Amplitude", fontsize=12)
plt.title("Fourier-Spektrum der Riemann-Zeta-Nullstellen", fontsize=14)
plt.legend()
plt.grid()
plt.show()

# üìå 4Ô∏è‚É£ Identifikation dominanter Frequenzen
dominant_frequencies = frequencies[np.argsort(spectrum)[-5:]]  # Die 5 st√§rksten Peaks
print(f"üîç Dominante Frequenzen der Nullstellen: {dominant_frequencies}")
```

```python
# üìå **Bereit f√ºr Google Colab**
import numpy as np
import requests
import matplotlib.pyplot as plt
import scipy.fftpack
import scipy.signal
import pywt
import re

# üìå **URLs der Nullstellendatens√§tze**
odlyzko_urls = {
    "zeros1 (100)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "zeros2 (10,000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "zeros6 (2,000,000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

lmfdb_url = "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1"

# üìå **Funktion: Nullstellen aus Odlyzko abrufen**
def load_odlyzko_zeros(url):
    print(f"üì• Lade Nullstellen aus {url} ...")
    response = requests.get(url)
    data = response.text.splitlines()
    zeros = [float(line.strip()) for line in data if line.strip().replace(".", "").isdigit()]
    print(f"‚úÖ {len(zeros)} Nullstellen geladen!")
    return np.array(zeros)

# üìå **Funktion: Nullstellen aus LMFDB abrufen & formatieren**
def load_lmfdb_zeros(url):
    print(f"üì• Lade 50.000 Nullstellen aus LMFDB ...")
    response = requests.get(url)
    data = response.text.splitlines()

    # Entferne Nummerierung ("1.; 2.; 3.;") aus jeder Zeile
    cleaned_zeros = []
    for line in data:
        match = re.search(r"[-+]?\d*\.\d+", line)  # Finde die erste Gleitkommazahl
        if match:
            cleaned_zeros.append(float(match.group(0)))

    print(f"‚úÖ {len(cleaned_zeros)} Nullstellen aus LMFDB erfolgreich geladen!")
    return np.array(cleaned_zeros)

# üìå **Funktion: Frequenzanalyse durchf√ºhren**
def frequency_analysis(zeros, dataset_name):
    print(f"\nüìä **Starte Frequenzanalyse f√ºr {dataset_name}**")

    # ‚úÖ **Berechne Nullstellen-Abst√§nde**
    abstaende = np.diff(zeros)
    abstaende -= np.mean(abstaende)  # Mittelwert abziehen (Hochpassfilter)

    # ‚úÖ **1Ô∏è‚É£ Fourier-Analyse (FFT)**
    spektrum_fft = np.abs(scipy.fftpack.fft(abstaende))
    frequenzen_fft = scipy.fftpack.fftfreq(len(abstaende), d=np.median(abstaende))
    dominante_frequenz_fft = abs(frequenzen_fft[np.argmax(spektrum_fft)])

    # ‚úÖ **2Ô∏è‚É£ Wavelet-Analyse**
    coeffs, freqs = pywt.cwt(abstaende, np.arange(1, 128), 'cmor1.5-1.0')
    dominante_frequenz_wavelet = freqs[np.argmax(np.abs(coeffs))]

    # ‚úÖ **3Ô∏è‚É£ Short-Time Fourier Transform (STFT)**
    f_stft, t_stft, Zxx = scipy.signal.spectrogram(abstaende, nperseg=5)
    dominante_frequenz_stft = f_stft[np.argmax(np.abs(Zxx))] if len(f_stft) > 0 else 0

    # ‚úÖ **4Ô∏è‚É£ Theoretischer Vergleichswert**
    phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
    theoretische_frequenz = (np.pi - phi) / np.pi

    # üìä **Ergebnisse ausgeben**
    print("\nüîç Vergleich der Ergebnisse:")
    print(f"‚úÖ Fourier-Analyse (FFT): {dominante_frequenz_fft:.6f}")
    print(f"üåä Wavelet-Analyse: {dominante_frequenz_wavelet:.6f}")
    print(f"üéµ Short-Time Fourier Transform (STFT): {dominante_frequenz_stft:.6f}")
    print(f"üìè Theoretischer Wert (œÄ - œÜ) / œÄ: {theoretische_frequenz:.6f}")

    # üìà **Abweichungen berechnen**
    abweichung_fft = abs(dominante_frequenz_fft - theoretische_frequenz)
    abweichung_wavelet = abs(dominante_frequenz_wavelet - theoretische_frequenz)
    abweichung_stft = abs(dominante_frequenz_stft - theoretische_frequenz)

    print("\nüìä Abweichungen vom theoretischen Wert:")
    print(f"üìå FFT-Abweichung: {abweichung_fft:.6f}")
    print(f"üìå Wavelet-Abweichung: {abweichung_wavelet:.6f}")
    print(f"üìå STFT-Abweichung: {abweichung_stft:.6f}")

    # üîé **Beste Methode w√§hlen**
    beste_methode = "FFT"
    min_abweichung = abweichung_fft
    if abweichung_wavelet < min_abweichung:
        beste_methode = "Wavelet"
        min_abweichung = abweichung_wavelet
    if abweichung_stft < min_abweichung:
        beste_methode = "STFT"

    print(f"\n‚úÖ Beste √úbereinstimmung mit Theorie: {beste_methode}")

    # üìä **Visualisierung**
    plt.figure(figsize=(8,6))
    plt.bar(["FFT", "Wavelet", "STFT", "Theorie"],
            [dominante_frequenz_fft, dominante_frequenz_wavelet, dominante_frequenz_stft, theoretische_frequenz],
            color=["blue", "green", "purple", "red"])
    plt.ylabel("Frequenz-Wert")
    plt.title(f"Vergleich der Frequenzmethoden ({dataset_name})")
    plt.ylim(0, 1)
    plt.show()

# ‚úÖ **Hauptprogramm: Lade & Analysiere alle Datens√§tze**
datasets = {}

# Odlyzko-Daten abrufen
for name, url in odlyzko_urls.items():
    datasets[name] = load_odlyzko_zeros(url)

# LMFDB-Daten abrufen
datasets["LMFDB (50,000)"] = load_lmfdb_zeros(lmfdb_url)

# Frequenzanalyse f√ºr alle Datens√§tze durchf√ºhren
for dataset_name, zeros in datasets.items():
    frequency_analysis(zeros, dataset_name)
```

```python
import numpy as np
import scipy.fftpack
import scipy.signal
import matplotlib.pyplot as plt

# ‚úÖ 1Ô∏è‚É£ Nullstellen-Daten (Testdaten)
nullstellen = np.array([
    14.134725142, 21.022039639, 25.01085758, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478
])

# ‚úÖ 2Ô∏è‚É£ Berechnung der Abst√§nde & logarithmischen Abst√§nde
abstaende = np.diff(nullstellen)
log_abstaende = np.log(abstaende)

# ‚úÖ 3Ô∏è‚É£ FFT auf log-Abst√§nden
spektrum = np.abs(scipy.fftpack.fft(log_abstaende))
frequenzen = scipy.fftpack.fftfreq(len(log_abstaende), d=np.median(log_abstaende))

# üîç Alternative Skalierung: Min-Max-Normalisierung (statt Z-Skalierung)
frequenzen_norm = (np.abs(frequenzen) - np.min(np.abs(frequenzen))) / (np.max(np.abs(frequenzen)) - np.min(np.abs(frequenzen)) + 1e-9)

# Dominante Frequenzen ausw√§hlen (Top 5, aber mit Median)
top_freqs = np.sort(np.abs(frequenzen_norm[np.argsort(spektrum)[-5:]]))
dominante_frequenz_median = np.median(top_freqs)

# ‚úÖ 4Ô∏è‚É£ Short-Time Fourier Transform (STFT)
f_stft, t_stft, Zxx = scipy.signal.spectrogram(log_abstaende, nperseg=5)
dominante_frequenz_stft = np.median(f_stft[np.argsort(np.abs(Zxx))[-5:]]) if len(f_stft) > 0 else 0

# ‚úÖ 5Ô∏è‚É£ Vergleich mit theoretischem Wert (œÄ - œÜ) / œÄ
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretische_frequenz = (np.pi - phi) / np.pi

# ‚úÖ 6Ô∏è‚É£ Visualisierung der Ergebnisse
plt.figure(figsize=(8,6))
plt.bar(["FFT (Min-Max)", "STFT", "Theoretischer Wert"],
        [dominante_frequenz_median, dominante_frequenz_stft, theoretische_frequenz],
        color=["blue", "green", "red"])
plt.ylabel("Frequenz-Wert")
plt.title("Vergleich der Frequenzen (Min-Max FFT vs. STFT)")
plt.ylim(0, 1)
plt.show()

# ‚úÖ 7Ô∏è‚É£ Ergebnisse ausgeben
print("\nüîç Vergleich der Ergebnisse:")
print(f"‚úÖ Median der Top 5 FFT-Frequenzen (Min-Max): {dominante_frequenz_median:.6f}")
print(f"üéµ Short-Time Fourier Transform (STFT): {dominante_frequenz_stft:.6f}")
print(f"üìè Theoretischer Wert (œÄ - œÜ) / œÄ: {theoretische_frequenz:.6f}")

# ‚úÖ 8Ô∏è‚É£ Abweichung berechnen
abweichung_fft = abs(dominante_frequenz_median - theoretische_frequenz)
abweichung_stft = abs(dominante_frequenz_stft - theoretische_frequenz)

print(f"\nüìä Abweichung vom theoretischen Wert:")
print(f"üìå FFT-Abweichung: {abweichung_fft:.6f}")
print(f"üìå STFT-Abweichung: {abweichung_stft:.6f}")

# ‚úÖ Beste Methode ermitteln
beste_methode = "FFT (Min-Max)" if abweichung_fft < abweichung_stft else "STFT"
print(f"\n‚úÖ Beste √úbereinstimmung mit Theorie: {beste_methode}")
```

```python
import numpy as np
import scipy.fftpack
import scipy.signal
import pywt
import matplotlib.pyplot as plt

# ‚úÖ 1Ô∏è‚É£ Nullstellen-Daten (Odlyzko)
nullstellen = np.array([
    14.134725142, 21.022039639, 25.01085758, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478
])

# ‚úÖ 2Ô∏è‚É£ Berechnung der Abst√§nde & logarithmischen Abst√§nde
abstaende = np.diff(nullstellen)
log_abstaende = np.log(abstaende)

# ‚úÖ 3Ô∏è‚É£ Normalisierung der Abst√§nde
min_max_abstaende = (abstaende - np.min(abstaende)) / (np.max(abstaende) - np.min(abstaende) + 1e-9)

# ‚úÖ 4Ô∏è‚É£ FFT mit verbesserter Fensterung (Blackman-Harris)
windowed_abstaende = min_max_abstaende * np.blackman(len(min_max_abstaende))
spektrum = np.abs(scipy.fftpack.fft(windowed_abstaende))
frequenzen = scipy.fftpack.fftfreq(len(windowed_abstaende), d=np.median(windowed_abstaende))

# üîç Frequenzextraktion: Median der Top 5 Frequenzen
top_freqs = np.sort(np.abs(frequenzen[np.argsort(spektrum)[-5:]]))
dominante_frequenz_fft = np.median(top_freqs)

# ‚úÖ 5Ô∏è‚É£ Optimierte Wavelet-Analyse (cmor 1.5-1.0)
coeffs, freqs = pywt.cwt(min_max_abstaende, np.arange(1, 128), 'cmor1.5-1.0')
dominante_frequenz_wavelet = np.median(freqs[np.argsort(np.abs(coeffs))[-5:]])

# ‚úÖ 6Ô∏è‚É£ Hybrid-Ansatz: Kombination von FFT & Wavelet
alpha = 0.5  # Gewichtungsfaktor anpassen
dominante_frequenz_hybrid = alpha * dominante_frequenz_fft + (1 - alpha) * dominante_frequenz_wavelet

# ‚úÖ 7Ô∏è‚É£ Theoretischer Vergleichswert (œÄ - œÜ) / œÄ
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretische_frequenz = (np.pi - phi) / np.pi

# ‚úÖ 8Ô∏è‚É£ Visualisierung der Ergebnisse
plt.figure(figsize=(8,6))
plt.bar(["FFT (Blackman-Harris)", "Wavelet (cmor)", "Hybrid", "Theoretischer Wert"],
        [dominante_frequenz_fft, dominante_frequenz_wavelet, dominante_frequenz_hybrid, theoretische_frequenz],
        color=["blue", "green", "purple", "red"])
plt.ylabel("Frequenz-Wert")
plt.title("Vergleich der Methoden zur Frequenzbestimmung")
plt.ylim(0, 1)
plt.show()

# ‚úÖ 9Ô∏è‚É£ Ergebnisse ausgeben
print("\nüîç Vergleich der Ergebnisse:")
print(f"‚úÖ FFT (Blackman-Harris): {dominante_frequenz_fft:.6f}")
print(f"üåä Wavelet-Analyse (cmor 1.5-1.0): {dominante_frequenz_wavelet:.6f}")
print(f"üîÑ Hybrid-Ansatz (FFT & Wavelet kombiniert): {dominante_frequenz_hybrid:.6f}")
print(f"üìè Theoretischer Wert (œÄ - œÜ) / œÄ: {theoretische_frequenz:.6f}")

# ‚úÖ üîé Abweichung berechnen
abweichung_fft = abs(dominante_frequenz_fft - theoretische_frequenz)
abweichung_wavelet = abs(dominante_frequenz_wavelet - theoretische_frequenz)
abweichung_hybrid = abs(dominante_frequenz_hybrid - theoretische_frequenz)

print(f"\nüìä Abweichung vom theoretischen Wert:")
print(f"üìå FFT-Abweichung: {abweichung_fft:.6f}")
print(f"üìå Wavelet-Abweichung: {abweichung_wavelet:.6f}")
print(f"üìå Hybrid-Abweichung: {abweichung_hybrid:.6f}")

# ‚úÖ Beste Methode bestimmen
beste_methode = min(
    ("FFT (Blackman-Harris)", abweichung_fft),
    ("Wavelet (cmor 1.5-1.0)", abweichung_wavelet),
    ("Hybrid", abweichung_hybrid),
    key=lambda x: x[1]
)[0]

print(f"\n‚úÖ Beste √úbereinstimmung mit Theorie: {beste_methode}")
```

```python
# Installiere ben√∂tigte Bibliotheken falls erforderlich (nur f√ºr Google Colab notwendig)
# !pip install numpy matplotlib scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import golden  # Der goldene Schnitt œÜ
import math

# Definiere die gegebenen Resonanzfrequenzen
frequencies = np.array([
    0.0667, 0.0801, 0.0994, 0.1087, 0.1121, 0.1167, 0.1194, 0.1214, 0.1241, 0.1254
])

# Gegebene Koh√§renzl√§ngen L(N) f√ºr verschiedene N-Werte
L_values = np.array([
    488.6906, 112.7362, 60745668422470623232.0, 125447844957281222656.0,
    101333078406337789952.0, 134936926060314705920.0, 118734002798476476416.0,
    91133807247113224192.0, 161202109847784357888.0, 120729313847970562048.0
])

# Berechnung der inversen Frequenzen
inverse_frequencies = 1 / frequencies[:len(L_values)]

# Berechnung des urspr√ºnglichen Skalierungsfaktors
scaling_factor = np.mean(L_values) / np.mean(inverse_frequencies)

# Berechnung der Korrekturterme
tau_ln2 = math.log(2) / (8 * math.pi)  # ln(2) / 8œÄ
tau_fibonacci = 1 / (golden**2 * math.pi)  # 1 / (œÜ¬≤ œÄ)

# Berechnung der verschiedenen korrigierten Skalierungsfaktoren
corrected_scaling_factor_ln2 = scaling_factor * (1 + tau_ln2)
corrected_scaling_factor_fibonacci = scaling_factor * (1 + tau_fibonacci)
corrected_scaling_factor_combined = scaling_factor * (1 + 0.5 * (tau_ln2 + tau_fibonacci))  # 50/50 Gewichtung

# Berechnung der gesch√§tzten Koh√§renzl√§ngen f√ºr jede Methode
L_estimated_original = scaling_factor * inverse_frequencies
L_estimated_ln2 = corrected_scaling_factor_ln2 * inverse_frequencies
L_estimated_fibonacci = corrected_scaling_factor_fibonacci * inverse_frequencies
L_estimated_combined = corrected_scaling_factor_combined * inverse_frequencies

# Visualisierung des Vergleichs
plt.figure(figsize=(10, 6))
plt.scatter(frequencies[:len(L_values)], L_values, color='black', label="Empirische Koh√§renzl√§nge (Messwerte)")
plt.plot(frequencies[:len(L_values)], L_estimated_original, 'b--', label="Sch√§tzung mit Original-Skalierungsfaktor")
plt.plot(frequencies[:len(L_values)], L_estimated_ln2, 'r-', label="Sch√§tzung mit Korrektur ln(2)/8œÄ")
plt.plot(frequencies[:len(L_values)], L_estimated_fibonacci, 'g-', label="Sch√§tzung mit Korrektur 1/(œÜ¬≤ œÄ)")
plt.plot(frequencies[:len(L_values)], L_estimated_combined, 'm-', label="Sch√§tzung mit kombinierter Korrektur")

plt.xlabel("Frequenz")
plt.ylabel("Koh√§renzl√§nge L(N)")
plt.title("Vergleich von berechneten und empirischen Koh√§renzl√§ngen")
plt.legend()
plt.grid(True)
plt.show()

# Berechnung der mittleren absoluten Abweichung f√ºr jede Methode
mae_original = np.mean(np.abs(L_values - L_estimated_original))
mae_ln2 = np.mean(np.abs(L_values - L_estimated_ln2))
mae_fibonacci = np.mean(np.abs(L_values - L_estimated_fibonacci))
mae_combined = np.mean(np.abs(L_values - L_estimated_combined))

# Ausgabe der Abweichungen
print("Mittlere absolute Abweichung (MAE) f√ºr jede Methode:")
print(f"üîπ Originaler Skalierungsfaktor: {mae_original:.5e}")
print(f"üîπ Mit Korrektur ln(2)/8œÄ: {mae_ln2:.5e}")
print(f"üîπ Mit Korrektur 1/(œÜ¬≤ œÄ): {mae_fibonacci:.5e}")
print(f"üîπ Mit kombinierter Korrektur: {mae_combined:.5e}")

# Bestimmung der besten Methode
best_method = min(mae_original, mae_ln2, mae_fibonacci, mae_combined)
if best_method == mae_original:
    best_name = "Originaler Skalierungsfaktor"
elif best_method == mae_ln2:
    best_name = "Mit Korrektur ln(2)/8œÄ"
elif best_method == mae_fibonacci:
    best_name = "Mit Korrektur 1/(œÜ¬≤ œÄ)"
else:
    best_name = "Mit kombinierter Korrektur"

print(f"\n‚úÖ Beste Methode zur Vorhersage der Koh√§renzl√§nge: {best_name}")
```
