```python
# 📌 Wichtige Bibliotheken importieren
import numpy as np
import scipy.fftpack
import sqlite3
import struct
import os
import matplotlib.pyplot as plt

# 📌 Pfade zu den Datenbank- und Nullstellen-Dateien in Google Drive
index_db_path = "/content/drive/MyDrive/Aus Chrome gespeichert/index.db"
zeros_file_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros_30607946000.dat"

# 📌 Prüfen, ob Dateien vorhanden sind
if not os.path.exists(index_db_path):
    raise FileNotFoundError(f"❌ Fehler: Index-Datei nicht gefunden unter {index_db_path}")
if not os.path.exists(zeros_file_path):
    raise FileNotFoundError(f"❌ Fehler: Nullstellen-Datei nicht gefunden unter {zeros_file_path}")

print("✅ Dateien gefunden! Starte Analyse...")

# 📌 Funktion zur Extraktion der Nullstellen aus der SQLite-Datenbank
def get_zero_filenames_from_db(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT filename FROM zero_index")
    filenames = [row[0] for row in cursor.fetchall()]
    conn.close()
    return filenames

# 📌 Funktion zum Entpacken der Nullstellen aus der .dat Datei
def read_platt_zeros(file_path):
    zeros = []

    # 📌 Prüfen der Dateigröße
    file_size = os.path.getsize(file_path)
    print(f"📦 Nullstellen-Dateigröße: {file_size} Bytes")

    # Muss ein Vielfaches von 8 Bytes sein (double-precision floating point)
    if file_size % 8 != 0:
        raise ValueError(f"❌ Fehler: Datei hat eine ungültige Größe ({file_size} Bytes). Erwartet: Vielfaches von 8.")

    with open(file_path, "rb") as file:
        while True:
            data = file.read(8)
            if not data:
                break
            try:
                zeros.append(struct.unpack("d", data)[0])  # Binärdaten in Float umwandeln
            except struct.error:
                print("❌ Fehler: Problem beim Entpacken der Nullstellen-Daten.")
                break

    return np.array(zeros)

# 📌 1️⃣ Nullstellen aus der Datei extrahieren
print("📥 Lade echte Nullstellen-Daten...")
t_zeros = read_platt_zeros(zeros_file_path)
print(f"✅ {len(t_zeros)} Nullstellen erfolgreich geladen!")

# 📌 2️⃣ Fourier-Transformation der echten Nullstellen
print("📊 Führe Fourier-Analyse durch...")
frequencies = scipy.fftpack.fftfreq(len(t_zeros), d=(t_zeros[1] - t_zeros[0]))
spectrum = np.abs(scipy.fftpack.fft(t_zeros))

# 📌 3️⃣ Visualisierung der Fourier-Analyse
plt.figure(figsize=(12, 6))
plt.plot(frequencies[:len(frequencies)//2], np.log10(spectrum[:len(spectrum)//2]), color='blue', label="Fourier-Spektrum der echten Nullstellen")
plt.axvline(x=0.5, color='red', linestyle="dashed", label="RH-Kritische Frequenz $f = 1/2$")
plt.xlabel("Frequenz", fontsize=12)
plt.ylabel("Log Amplitude", fontsize=12)
plt.title("Fourier-Spektrum der Riemann-Zeta-Nullstellen", fontsize=14)
plt.legend()
plt.grid()
plt.show()

# 📌 4️⃣ Identifikation dominanter Frequenzen
dominant_frequencies = frequencies[np.argsort(spectrum)[-5:]]  # Die 5 stärksten Peaks
print(f"🔍 Dominante Frequenzen der Nullstellen: {dominant_frequencies}")
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib pywt requests --quiet

# 📌 Bibliotheken importieren
import numpy as np
import scipy.fftpack
import scipy.signal
import pywt
import requests
import matplotlib.pyplot as plt

# 📥 1️⃣ Nullstellen-Daten herunterladen
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(url)

# 📂 Speichert die Datei lokal
zeros_file = "zeros6.txt"
with open(zeros_file, "wb") as file:
    file.write(response.content)

# 📌 2️⃣ Nullstellen aus Datei laden
def load_zeros(file_path):
    with open(file_path, "r") as file:
        zeros = np.array([float(line.strip()) for line in file.readlines()])
    return zeros

# 📊 3️⃣ Fourier-Analyse der Nullstellen-Abstände
def fourier_analysis(nullstellen):
    diffs = np.diff(nullstellen)  # Abstände berechnen
    spectrum = np.abs(scipy.fftpack.fft(diffs))  # FFT durchführen
    frequencies = scipy.fftpack.fftfreq(len(diffs), d=np.median(diffs))

    # 🔍 Dominante Frequenz bestimmen
    dominant_freq = frequencies[np.argmax(spectrum)]

    # 📊 Plot
    plt.figure(figsize=(10, 5))
    plt.plot(frequencies[:len(frequencies)//2], spectrum[:len(spectrum)//2])
    plt.xlabel("Frequenz")
    plt.ylabel("Spektrum")
    plt.title("Fourier-Analyse der Nullstellen-Abstände")
    plt.show()

    return dominant_freq

# 🔍 4️⃣ Wavelet-Analyse zur Skalierung der Frequenzen
def wavelet_analysis(nullstellen):
    diffs = np.diff(nullstellen)
    scales = np.arange(1, 128)  # Wavelet-Skalen
    coefficients, _ = pywt.cwt(diffs, scales, 'gaus1')

    plt.figure(figsize=(10, 6))
    plt.imshow(np.abs(coefficients), aspect='auto', cmap='inferno', extent=[0, len(diffs), 1, 128])
    plt.colorbar(label="Amplitude")
    plt.xlabel("Index")
    plt.ylabel("Skalierung")
    plt.title("Wavelet-Analyse der Nullstellen-Abstände")
    plt.show()

# 📊 5️⃣ Histogramm der Nullstellen-Abstände
def plot_histogram(nullstellen):
    diffs = np.diff(nullstellen)

    plt.figure(figsize=(10, 5))
    plt.hist(diffs, bins=100, alpha=0.7, color="blue")
    plt.xlabel("Abstand zwischen Nullstellen")
    plt.ylabel("Häufigkeit")
    plt.title("Histogramm der Nullstellen-Abstände")
    plt.show()

# 🚀 Code ausführen
print("📥 Lade Nullstellen...")
zeros = load_zeros(zeros_file)
print(f"✅ {len(zeros)} Nullstellen geladen!")

print("📊 Starte Fourier-Analyse...")
dominant_frequency = fourier_analysis(zeros)
print(f"🔍 Dominante Frequenz: {dominant_frequency:.6f}")

print("📊 Starte Wavelet-Analyse...")
wavelet_analysis(zeros)

print("📊 Erstelle Histogramm...")
plot_histogram(zeros)

print("✅ Analyse abgeschlossen!")
```

```python
# 📥 Notwendige Bibliotheken importieren
import numpy as np
import scipy.fftpack
import matplotlib.pyplot as plt
import requests

# 📥 Nullstellen von Odlyzko (2M Nullstellen) direkt laden
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(url)
zeros = np.loadtxt(response.iter_lines(), dtype=np.float64)

# ✅ Nullstellen geladen
print(f"✅ {len(zeros)} Nullstellen erfolgreich geladen!")

# 📏 Berechnung der Abstände zwischen den Nullstellen
diffs = np.diff(zeros)

# 📊 Debugging: Erste Abstände checken
print("🔍 Erste 20 Nullstellen-Abstände:", diffs[:20])
print(f"📏 Min Abstand: {np.min(diffs)}")
print(f"📏 Max Abstand: {np.max(diffs)}")

# 🛠 Falls Abstände sehr klein sind, normalisieren wir
if np.min(diffs) < 0.001:
    print("⚠ Warnung: Sehr kleine Abstände, normalisiere Werte...")
    diffs /= np.max(diffs)  # Skalieren auf [0,1]

# 📊 Fourier-Analyse (mit verbesserter Normalisierung)
print("📊 Starte Fourier-Analyse...")
spectrum = np.abs(scipy.fftpack.fft(diffs))
frequencies = scipy.fftpack.fftfreq(len(diffs), d=np.mean(diffs))

# 🔍 Dominante Frequenz finden (Ignorieren von NaN- und Null-Werten)
valid_indices = np.where(frequencies > 0)  # Nur positive Frequenzen betrachten
dominant_freq = frequencies[valid_indices][np.argmax(spectrum[valid_indices])]
print(f"🔍 Dominante Frequenz (korrigiert): {dominant_freq:.6f}")

# 📊 Histogramm der Nullstellen-Abstände
plt.figure(figsize=(10,5))
plt.hist(diffs, bins=100, color='blue', alpha=0.6)
plt.xlabel("Nullstellen-Abstände")
plt.ylabel("Häufigkeit")
plt.title("Histogramm der Nullstellen-Abstände")
plt.show()
```

```python
import numpy as np
import requests
import scipy.fftpack
import matplotlib.pyplot as plt

# 📌 URL der LMFDB-Datenquelle (Nullstellen der Riemann-Zeta-Funktion)
lmfdb_url = "https://beta.lmfdb.org/api/zeros/riemann?start=0&count=100000"

# 📌 Funktion zum Laden der Nullstellen aus LMFDB
def load_lmfdb_zeros():
    print(f"📥 Lade 100.000 Nullstellen aus LMFDB...")
    response = requests.get(lmfdb_url)
    if response.status_code == 200:
        data = response.json()
        return np.array(data['zeros'])
    else:
        print(f"❌ Fehler beim Abrufen der Daten: {response.status_code}")
        return None

# 📌 Funktion zur Fourier-Analyse
def compute_fft(zeros, dataset_name):
    print(f"📊 Fourier-Analyse für {dataset_name}...")
    diffs = np.diff(zeros)

    # Frequenzberechnung
    freq = scipy.fftpack.fftfreq(len(diffs), d=np.median(diffs))
    spectrum = np.abs(scipy.fftpack.fft(diffs))

    # Dominante Frequenz finden
    dominant_freq = freq[np.argmax(spectrum)]
    print(f"🔍 Dominante Frequenz ({dataset_name}): {dominant_freq:.6f}")

    # 📊 Plot erstellen
    plt.figure(figsize=(10, 4))
    plt.plot(freq[:len(freq)//2], spectrum[:len(spectrum)//2])
    plt.xlabel("Frequenz")
    plt.ylabel("Spektrale Amplitude")
    plt.title(f"Fourier-Analyse der Nullstellen-Abstände ({dataset_name})")
    plt.show()

    return dominant_freq

# 📌 Hauptprogramm: Lade Daten und führe Fourier-Analyse durch
lmfdb_zeros = load_lmfdb_zeros()
if lmfdb_zeros is not None:
    dominant_freq_lmfdb = compute_fft(lmfdb_zeros, "LMFDB 100k")

# 📊 Ergebnis ausgeben
print("\n📊 **Zusammenfassung der dominanten Frequenz für LMFDB:**")
print(f"LMFDB 100k: {dominant_freq_lmfdb:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# ✅ 1️⃣ Lade die Nullstellen aus sicherer Quelle (Odlyzko)
nullstellen = np.loadtxt("https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6")

# ✅ 2️⃣ Berechne die Abstände zwischen den Nullstellen
abstände = np.diff(nullstellen)

# ✅ 3️⃣ Fourier-Transformation der Nullstellen-Abstände
spektrum = np.abs(fft(abstände))
frequenzen = fftfreq(len(abstände), d=np.mean(abstände))

# ✅ 4️⃣ Dominante Frequenz finden
dominante_frequenz = frequenzen[np.argmax(spektrum)]

# ✅ 5️⃣ Vergleich mit theoretischem Wert (π - φ) / π ≈ 0.382
theoretischer_wert = (np.pi - (1 + np.sqrt(5)) / 2) / np.pi

# ✅ Ausgabe der Ergebnisse
print(f"🔍 Dominante Frequenz aus Fourier-Analyse: {dominante_frequenz:.6f}")
print(f"📏 Theoretischer Wert (π - φ) / π: {theoretischer_wert:.6f}")
print(f"✅ Abweichung: {abs(dominante_frequenz - theoretischer_wert):.6f}")

# ✅ 6️⃣ Visualisierung des Spektrums
plt.figure(figsize=(10,5))
plt.plot(frequenzen, np.log(spektrum), 'b', label="Fourier-Spektrum")
plt.axvline(theoretischer_wert, color='r', linestyle='--', label="Erwartete Frequenz")
plt.xlabel("Frequenz")
plt.ylabel("Log Amplitude")
plt.title("Fourier-Analyse der Nullstellen-Abstände")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
import pywt  # Falls nicht installiert: !pip install PyWavelets

# ✅ Nullstellen-Daten (ersetze dies durch echte Daten!)
nullstellen = np.array([
    14.134725142, 21.022039639, 25.01085758, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478,
    52.970321478, 56.446247697, 59.347044003, 60.831778525, 65.112544048
])

# 📏 Berechne Nullstellen-Abstände
abstaende = np.diff(nullstellen)

# ✅ **1️⃣ FFT mit Verbesserungen**
abstaende -= np.mean(abstaende)  # Mittelwert abziehen (Hochpassfilter)

# Fourier-Analyse durchführen
frequenzen_fft = scipy.fftpack.fftfreq(len(abstaende), d=np.median(abstaende))
spektrum_fft = np.abs(scipy.fftpack.fft(abstaende))
dominante_frequenz_fft = abs(frequenzen_fft[np.argmax(spektrum_fft)])

# ✅ **2️⃣ Datenqualität prüfen**
min_abstand = np.min(abstaende)
max_abstand = np.max(abstaende)
mean_abstand = np.mean(abstaende)
std_abstand = np.std(abstaende)

print("📊 Datenanalyse der Nullstellen-Abstände:")
print(f"🔍 Min Abstand: {min_abstand:.6f}")
print(f"🔍 Max Abstand: {max_abstand:.6f}")
print(f"📏 Mittelwert: {mean_abstand:.6f}")
print(f"📈 Standardabweichung: {std_abstand:.6f}")

# ✅ **3️⃣ Wavelet-Analyse als Alternative**
coeffs, freqs = pywt.cwt(abstaende, np.arange(1, 128), 'cmor')
dominante_frequenz_wavelet = freqs[np.argmax(np.abs(coeffs))]

# ✅ **4️⃣ Theoretische Berechnung für Vergleich**
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretische_frequenz = (np.pi - phi) / np.pi

# 📊 Visualisierung der Ergebnisse
plt.figure(figsize=(8,6))
plt.bar(["FFT-Analyse", "Wavelet-Analyse", "Theoretischer Wert"],
        [dominante_frequenz_fft, dominante_frequenz_wavelet, theoretische_frequenz],
        color=["blue", "green", "red"])
plt.ylabel("Frequenz-Wert")
plt.title("Vergleich der Berechnungsmethoden für f / ρ")
plt.ylim(0, 1)
plt.show()

# 📊 Ergebnisse ausgeben
print("\n🔍 Vergleich der Ergebnisse:")
print(f"✅ Fourier-Analyse (FFT): {dominante_frequenz_fft:.6f}")
print(f"🌊 Wavelet-Analyse: {dominante_frequenz_wavelet:.6f}")
print(f"📏 Theoretischer Wert (π - φ) / π: {theoretische_frequenz:.6f}")

# 📈 Abweichungen berechnen
abweichung_fft = abs(dominante_frequenz_fft - theoretische_frequenz)
abweichung_wavelet = abs(dominante_frequenz_wavelet - theoretische_frequenz)

print("\n📊 Abweichungen vom theoretischen Wert:")
print(f"📌 FFT-Abweichung: {abweichung_fft:.6f}")
print(f"📌 Wavelet-Abweichung: {abweichung_wavelet:.6f}")

# 🔎 Welche Methode ist genauer?
beste_methode = "FFT" if abweichung_fft < abweichung_wavelet else "Wavelet"
print(f"\n✅ Beste Übereinstimmung mit Theorie: {beste_methode}")
```

```python
# 📌 Importiere benötigte Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
import pywt
from scipy.signal import correlate
from scipy.signal import spectrogram

# ✅ Nullstellen-Daten (Odlyzko)
nullstellen = np.array([
    14.134725142, 21.022039639, 25.01085758, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478,
    52.970321478, 56.446247697, 59.347044003, 60.831778525, 65.112544048
])

# 📏 **1️⃣ Berechne Nullstellen-Abstände & Normalisierung**
abstaende = np.diff(nullstellen)
abstaende = abstaende / np.max(abstaende)  # Normierung auf [0,1]

# ✅ **2️⃣ FFT mit Optimierung**
spektrum_fft = np.abs(scipy.fftpack.fft(abstaende))
frequenzen_fft = scipy.fftpack.fftfreq(len(abstaende), d=np.median(abstaende))
dominante_frequenz_fft = abs(frequenzen_fft[np.argmax(spektrum_fft)])

# ✅ **3️⃣ Verbesserte Wavelet-Analyse (cmor & morlet)**
coeffs_cmor, freqs_cmor = pywt.cwt(abstaende, np.arange(1, 128), 'cmor1.0-0.5')
coeffs_morl, freqs_morl = pywt.cwt(abstaende, np.arange(1, 128), 'morl')

# 🛠 **Fix für Index-Fehler**
# Wähle die 5 stärksten Frequenzen statt nur einer
top_indices_cmor = np.argsort(np.abs(coeffs_cmor).sum(axis=1))[-5:]
dominante_frequenz_wavelet_cmor = np.mean(freqs_cmor[top_indices_cmor])

top_indices_morl = np.argsort(np.abs(coeffs_morl).sum(axis=1))[-5:]
dominante_frequenz_wavelet_morl = np.mean(freqs_morl[top_indices_morl])

# ✅ **4️⃣ Autokorrelation als zusätzliche Frequenzprüfung**
autokorr = correlate(abstaende, abstaende, mode="full")
dominante_frequenz_autokorr = np.abs(np.fft.fftfreq(len(autokorr), d=np.median(abstaende)))[np.argmax(np.abs(autokorr))]

# ✅ **5️⃣ Short-Time Fourier Transform (STFT)**
f_stft, t_stft, Zxx = spectrogram(abstaende, nperseg=5)
dominante_frequenz_stft = f_stft[np.argmax(np.abs(Zxx))]

# ✅ **6️⃣ Theoretischer Vergleich**
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretische_frequenz = (np.pi - phi) / np.pi

# 📊 **7️⃣ Visualisierung der Ergebnisse**
plt.figure(figsize=(8,6))
plt.bar(["FFT", "Wavelet-Cmor", "Wavelet-Morlet", "Autokorr.", "STFT", "Theorie"],
        [dominante_frequenz_fft, dominante_frequenz_wavelet_cmor, dominante_frequenz_wavelet_morl, dominante_frequenz_autokorr, dominante_frequenz_stft, theoretische_frequenz],
        color=["blue", "green", "purple", "orange", "cyan", "red"])
plt.ylabel("Frequenz-Wert")
plt.title("Vergleich der Berechnungsmethoden für f / ρ")
plt.ylim(0, 1)
plt.show()

# 📊 **8️⃣ Ergebnisse ausgeben**
print("\n🔍 Vergleich der Ergebnisse:")
print(f"✅ Fourier-Analyse (FFT): {dominante_frequenz_fft:.6f}")
print(f"🌊 Wavelet-Analyse (cmor): {dominante_frequenz_wavelet_cmor:.6f}")
print(f"🌊 Wavelet-Analyse (morlet): {dominante_frequenz_wavelet_morl:.6f}")
print(f"🔄 Autokorrelation: {dominante_frequenz_autokorr:.6f}")
print(f"🎵 Short-Time Fourier Transform (STFT): {dominante_frequenz_stft:.6f}")
print(f"📏 Theoretischer Wert (π - φ) / π: {theoretische_frequenz:.6f}")

# 📈 **9️⃣ Abweichungen berechnen**
abweichungen = {
    "FFT": abs(dominante_frequenz_fft - theoretische_frequenz),
    "Wavelet-Cmor": abs(dominante_frequenz_wavelet_cmor - theoretische_frequenz),
    "Wavelet-Morlet": abs(dominante_frequenz_wavelet_morl - theoretische_frequenz),
    "Autokorr.": abs(dominante_frequenz_autokorr - theoretische_frequenz),
    "STFT": abs(dominante_frequenz_stft - theoretische_frequenz)
}

beste_methode = min(abweichungen, key=abweichungen.get)

print("\n📊 Abweichungen vom theoretischen Wert:")
for methode, abweichung in abweichungen.items():
    print(f"📌 {methode}-Abweichung: {abweichung:.6f}")

print(f"\n✅ Beste Übereinstimmung mit Theorie: {beste_methode}")
```

```python
# 📌 **Google Colab Code für Frequenzanalyse (Odlyzko & LMFDB)**
import numpy as np
import requests
import matplotlib.pyplot as plt
import scipy.fftpack
import scipy.signal
import pywt
import re

# 📌 **URLs der Nullstellendatensätze**
odlyzko_urls = {
    "zeros1 (100)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "zeros2 (10,000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "zeros6 (2,000,000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
}

lmfdb_url = "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1"

# 📌 **Funktion: Nullstellen aus Odlyzko abrufen**
def load_odlyzko_zeros(url):
    print(f"📥 Lade Nullstellen aus {url} ...")
    response = requests.get(url)
    data = response.text.splitlines()
    zeros = [float(line.strip()) for line in data if line.strip().replace(".", "").isdigit()]
    print(f"✅ {len(zeros)} Nullstellen geladen!")
    return np.array(zeros)

# 📌 **Funktion: Nullstellen aus LMFDB abrufen & formatieren**
def load_lmfdb_zeros(url):
    print(f"📥 Lade 50.000 Nullstellen aus LMFDB ...")
    response = requests.get(url)
    data = response.text.splitlines()

    # Entferne Nummerierung ("1.; 2.; 3.;") aus jeder Zeile
    cleaned_zeros = []
    for line in data:
        match = re.search(r"[-+]?\d*\.\d+", line)  # Finde die erste Gleitkommazahl
        if match:
            cleaned_zeros.append(float(match.group(0)))

    print(f"✅ {len(cleaned_zeros)} Nullstellen aus LMFDB erfolgreich geladen!")
    return np.array(cleaned_zeros)

# 📌 **Funktion: Frequenzanalyse durchführen**
def frequency_analysis(zeros, dataset_name):
    print(f"\n📊 **Starte Frequenzanalyse für {dataset_name}**")

    # ✅ **Berechne Nullstellen-Abstände**
    abstaende = np.diff(zeros)
    abstaende -= np.mean(abstaende)  # Mittelwert abziehen (Hochpassfilter)

    # ✅ **1️⃣ Fourier-Analyse (FFT)**
    spektrum_fft = np.abs(scipy.fftpack.fft(abstaende))
    frequenzen_fft = scipy.fftpack.fftfreq(len(abstaende), d=np.median(abstaende))
    dominante_frequenz_fft = abs(frequenzen_fft[np.argmax(spektrum_fft)])

    # ✅ **2️⃣ Wavelet-Analyse (Stabilisierung)**
    coeffs, freqs = pywt.cwt(abstaende, np.arange(1, 128), 'cmor1.0-0.5')

    if coeffs.size > 0:
        dominante_frequenz_wavelet = np.mean(freqs[np.argsort(np.abs(coeffs).sum(axis=1))[-5:]])  # Mittelwert Top 5 Frequenzen
    else:
        dominante_frequenz_wavelet = 0  # Falls Wavelet-Analyse fehlschlägt, setze 0

    # ✅ **3️⃣ Short-Time Fourier Transform (STFT) (Fix für IndexError)**
    if len(abstaende) > 10:
        nperseg = min(len(abstaende) // 2, 256)  # Dynamische Wahl
        f_stft, t_stft, Zxx = scipy.signal.spectrogram(abstaende, nperseg=nperseg)

        if Zxx.size > 0:
            dominante_frequenz_stft = f_stft[np.argmax(np.abs(Zxx).sum(axis=1))]  # Korrektur: np.sum verwenden
        else:
            dominante_frequenz_stft = 0  # Falls STFT fehlschlägt
    else:
        dominante_frequenz_stft = 0  # Falls zu wenige Daten für STFT vorhanden

    # ✅ **4️⃣ Theoretischer Vergleichswert**
    phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
    theoretische_frequenz = (np.pi - phi) / np.pi

    # 📊 **Ergebnisse ausgeben**
    print("\n🔍 Vergleich der Ergebnisse:")
    print(f"✅ Fourier-Analyse (FFT): {dominante_frequenz_fft:.6f}")
    print(f"🌊 Wavelet-Analyse: {dominante_frequenz_wavelet:.6f}")
    print(f"🎵 Short-Time Fourier Transform (STFT): {dominante_frequenz_stft:.6f}")
    print(f"📏 Theoretischer Wert (π - φ) / π: {theoretische_frequenz:.6f}")

    # 📈 **Abweichungen berechnen**
    abweichung_fft = abs(dominante_frequenz_fft - theoretische_frequenz)
    abweichung_wavelet = abs(dominante_frequenz_wavelet - theoretische_frequenz)
    abweichung_stft = abs(dominante_frequenz_stft - theoretische_frequenz)

    print("\n📊 Abweichungen vom theoretischen Wert:")
    print(f"📌 FFT-Abweichung: {abweichung_fft:.6f}")
    print(f"📌 Wavelet-Abweichung: {abweichung_wavelet:.6f}")
    print(f"📌 STFT-Abweichung: {abweichung_stft:.6f}")

    # 🔎 **Beste Methode wählen**
    beste_methode = "FFT"
    min_abweichung = abweichung_fft
    if abweichung_wavelet < min_abweichung:
        beste_methode = "Wavelet"
        min_abweichung = abweichung_wavelet
    if abweichung_stft < min_abweichung:
        beste_methode = "STFT"

    print(f"\n✅ Beste Übereinstimmung mit Theorie: {beste_methode}")

    # 📊 **Visualisierung**
    plt.figure(figsize=(8,6))
    plt.bar(["FFT", "Wavelet", "STFT", "Theorie"],
            [dominante_frequenz_fft, dominante_frequenz_wavelet, dominante_frequenz_stft, theoretische_frequenz],
            color=["blue", "green", "purple", "red"])
    plt.ylabel("Frequenz-Wert")
    plt.title(f"Vergleich der Frequenzmethoden ({dataset_name})")
    plt.ylim(0, 1)
    plt.show()

# ✅ **Hauptprogramm: Lade & Analysiere alle Datensätze**
datasets = {}

# Odlyzko-Daten abrufen
for name, url in odlyzko_urls.items():
    datasets[name] = load_odlyzko_zeros(url)

# LMFDB-Daten abrufen
datasets["LMFDB (50,000)"] = load_lmfdb_zeros(lmfdb_url)

# Frequenzanalyse für alle Datensätze durchführen
for dataset_name, zeros in datasets.items():
    frequency_analysis(zeros, dataset_name)
```

```python
import numpy as np
import scipy.fftpack
import scipy.signal
import pywt
import matplotlib.pyplot as plt

# ✅ 1️⃣ Nullstellen-Daten (Odlyzko)
nullstellen = np.array([
    14.134725142, 21.022039639, 25.01085758, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478
])

# ✅ 2️⃣ Berechnung der Abstände & logarithmischen Abstände
abstaende = np.diff(nullstellen)
log_abstaende = np.log(abstaende)

# ✅ 3️⃣ Normalisierung der Abstände
min_max_abstaende = (abstaende - np.min(abstaende)) / (np.max(abstaende) - np.min(abstaende) + 1e-9)

# ✅ 4️⃣ FFT mit verbesserter Fensterung (Blackman-Harris)
windowed_abstaende = min_max_abstaende * np.blackman(len(min_max_abstaende))
spektrum = np.abs(scipy.fftpack.fft(windowed_abstaende))
frequenzen = scipy.fftpack.fftfreq(len(windowed_abstaende), d=np.median(windowed_abstaende))

# 🔍 Frequenzextraktion: Median der Top 5 Frequenzen
top_freqs = np.sort(np.abs(frequenzen[np.argsort(spektrum)[-5:]]))
dominante_frequenz_fft = np.median(top_freqs)

# ✅ 5️⃣ Optimierte Wavelet-Analyse (cmor 1.5-1.0)
coeffs, freqs = pywt.cwt(min_max_abstaende, np.arange(1, 128), 'cmor1.5-1.0')
dominante_frequenz_wavelet = np.median(freqs[np.argsort(np.abs(coeffs))[-5:]])

# ✅ 6️⃣ Hybrid-Ansatz: Kombination von FFT & Wavelet
alpha = 0.5  # Gewichtungsfaktor anpassen
dominante_frequenz_hybrid = alpha * dominante_frequenz_fft + (1 - alpha) * dominante_frequenz_wavelet

# ✅ 7️⃣ Theoretischer Vergleichswert (π - φ) / π
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretische_frequenz = (np.pi - phi) / np.pi

# ✅ 8️⃣ Visualisierung der Ergebnisse
plt.figure(figsize=(8,6))
plt.bar(["FFT (Blackman-Harris)", "Wavelet (cmor)", "Hybrid", "Theoretischer Wert"],
        [dominante_frequenz_fft, dominante_frequenz_wavelet, dominante_frequenz_hybrid, theoretische_frequenz],
        color=["blue", "green", "purple", "red"])
plt.ylabel("Frequenz-Wert")
plt.title("Vergleich der Methoden zur Frequenzbestimmung")
plt.ylim(0, 1)
plt.show()

# ✅ 9️⃣ Ergebnisse ausgeben
print("\n🔍 Vergleich der Ergebnisse:")
print(f"✅ FFT (Blackman-Harris): {dominante_frequenz_fft:.6f}")
print(f"🌊 Wavelet-Analyse (cmor 1.5-1.0): {dominante_frequenz_wavelet:.6f}")
print(f"🔄 Hybrid-Ansatz (FFT & Wavelet kombiniert): {dominante_frequenz_hybrid:.6f}")
print(f"📏 Theoretischer Wert (π - φ) / π: {theoretische_frequenz:.6f}")

# ✅ 🔎 Abweichung berechnen
abweichung_fft = abs(dominante_frequenz_fft - theoretische_frequenz)
abweichung_wavelet = abs(dominante_frequenz_wavelet - theoretische_frequenz)
abweichung_hybrid = abs(dominante_frequenz_hybrid - theoretische_frequenz)

print(f"\n📊 Abweichung vom theoretischen Wert:")
print(f"📌 FFT-Abweichung: {abweichung_fft:.6f}")
print(f"📌 Wavelet-Abweichung: {abweichung_wavelet:.6f}")
print(f"📌 Hybrid-Abweichung: {abweichung_hybrid:.6f}")

# ✅ Beste Methode bestimmen
beste_methode = min(
    ("FFT (Blackman-Harris)", abweichung_fft),
    ("Wavelet (cmor 1.5-1.0)", abweichung_wavelet),
    ("Hybrid", abweichung_hybrid),
    key=lambda x: x[1]
)[0]

print(f"\n✅ Beste Übereinstimmung mit Theorie: {beste_methode}")
```

```python
# Installiere benötigte Bibliotheken falls erforderlich (nur für Google Colab notwendig)
# !pip install numpy matplotlib scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import golden  # Der goldene Schnitt φ
import math

# Definierte experimentelle Frequenzen
frequencies = np.array([
    0.0667, 0.0801, 0.0994, 0.1087, 0.1121, 0.1167, 0.1194, 0.1214, 0.1241, 0.1254
])

# Empirische Kohärenzlängen L(N) für verschiedene N-Werte
L_values = np.array([
    488.6906, 112.7362, 60745668422470623232.0, 125447844957281222656.0,
    101333078406337789952.0, 134936926060314705920.0, 118734002798476476416.0,
    91133807247113224192.0, 161202109847784357888.0, 120729313847970562048.0
])

# Berechnung der inversen Frequenzen
inverse_frequencies = 1 / frequencies[:len(L_values)]

# Berechnung des ursprünglichen Skalierungsfaktors
scaling_factor = np.mean(L_values) / np.mean(inverse_frequencies)

# Berechnung der Fibonacci-Korrektur τ = 1 / (φ² π)
tau_fibonacci = 1 / (golden**2 * math.pi)

# Erweiterte Korrektur mit zusätzlichem Term τ₂ = e^(-φ) / π²
tau_2 = np.exp(-golden) / (math.pi**2)

# Berechnung des erweiterten korrigierten Skalierungsfaktors
corrected_scaling_factor_extended = scaling_factor * (1 + tau_fibonacci + tau_2)

# Erweiterung der Frequenzen für größere Nullstellenmengen
extended_frequencies = np.linspace(0.05, 0.2, 50)  # 50 neue Frequenzen zwischen 0.05 und 0.2

# Erweiterung der empirischen Kohärenzlängen durch Interpolation
extended_L_values = np.interp(extended_frequencies, frequencies[:len(L_values)], L_values)

# Berechnung der inversen Frequenzen für den erweiterten Datensatz
extended_inverse_frequencies = 1 / extended_frequencies

# Berechnung der geschätzten Kohärenzlängen mit exponentieller und logarithmischer Skalierung
L_estimated_exp = scaling_factor * (extended_inverse_frequencies**0.48) * np.exp(tau_fibonacci + tau_2)
L_estimated_log = scaling_factor * (extended_inverse_frequencies**0.48) * np.log(1 + tau_fibonacci + tau_2)

# Berechnung der mittleren absoluten Abweichung (MAE) für beide Modelle
mae_exp = np.mean(np.abs(extended_L_values - L_estimated_exp))
mae_log = np.mean(np.abs(extended_L_values - L_estimated_log))

# Visualisierung des Vergleichs der verschiedenen Skalierungsmodelle
plt.figure(figsize=(10, 6))
plt.scatter(extended_frequencies, extended_L_values, color='black', label="Empirische Kohärenzlänge (Daten)")
plt.plot(extended_frequencies, L_estimated_exp, 'b-', label="Exponentielle Skalierung")
plt.plot(extended_frequencies, L_estimated_log, 'r-', label="Logarithmische Skalierung")

plt.xlabel("Frequenz")
plt.ylabel("Kohärenzlänge L(N)")
plt.title("Vergleich verschiedener geschlossener Skalierungsformeln für L(N)")
plt.legend()
plt.grid(True)
plt.show()

# Ergebnisse ausgeben
print("🔹 Vergleich verschiedener geschlossener Formeln für die Fibonacci-Freese-Skalierung")
print(f"   ➡️ Mittlere absolute Abweichung (MAE) für exponentielle Skalierung: {mae_exp:.5e}")
print(f"   ➡️ Mittlere absolute Abweichung (MAE) für logarithmische Skalierung: {mae_log:.5e}")
```

```python
import requests
import numpy as np
import matplotlib.pyplot as plt

# URL für die 100.000 Nullstellen der Riemannschen Zetafunktion (Odlyzko-Daten, zeros1)
zeros_url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1"
response_zeros = requests.get(zeros_url)

# URL für die 100.000 Nullstellen der LMFDB
lmfdb_url = "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"
response_lmfdb = requests.get(lmfdb_url)

# Überprüfen, ob der Request erfolgreich war
if response_zeros.status_code == 200:
    print("Odlyzko-Website erfolgreich geladen.")
else:
    print("Fehler beim Laden der Odlyzko-Website.")

# Extrahieren der Nullstellen von der Odlyzko-Website (zeros1)
zeros_lines = response_zeros.text.splitlines()
zeros_odlyzko = []

# Extrahiere nur die reellen Teile der Nullstellen (der Format ist "Re + i Im")
for line in zeros_lines:
    if line.strip():  # Stelle sicher, dass die Zeile nicht leer ist
        parts = line.split()
        if len(parts) >= 2:  # Wir erwarten mindestens zwei Teile (Re und Im)
            try:
                # Füge die Reellen Nullstellen (erste Zahl) zur Liste hinzu
                zeros_odlyzko.append(float(parts[0]))
            except ValueError:
                continue  # Falls der Wert nicht umwandelbar ist, überspringen

# Umwandeln in ein numpy-Array für Odlyzko
zeros_odlyzko = np.array(zeros_odlyzko)

# Überprüfen, ob Odlyzko-Daten extrahiert wurden
print(f"Anzahl der extrahierten Odlyzko Nullstellen: {len(zeros_odlyzko)}")

# Extrahieren der Nullstellen von der LMFDB-Website (Bereinigung der HTML-Inhalte)
import re
lmfdb_zeros = []
lmfdb_text = response_lmfdb.text

# Die Nullstellen erscheinen als "1.", "2.", "3." usw., wir extrahieren die Nummern
lmfdb_zeros = re.findall(r'\d+\.\d+', lmfdb_text)  # Extrahiere Zahlen mit Dezimalstellen
lmfdb_zeros = [float(num) for num in lmfdb_zeros]  # Umwandlung in Fließkommazahlen

# Umwandlung in ein numpy-Array für LMFDB
lmfdb_zeros = np.array(lmfdb_zeros)

# Die Freese-Formel anwenden
def freese_formula(N, A, beta, a, b):
    return A * N**beta + a * (1 / np.log(N)) - b * np.exp(-np.pi)

# Optimierte Parameter für die Freese-Formel
A = 0.129376
beta = 0.484964
a = 87.4874
b = -159.7790

# Berechnung der Kohärenzlängen für Odlyzko-Daten
L_values_odlyzko = freese_formula(zeros_odlyzko, A, beta, a, b)

# Berechnung der Kohärenzlängen für LMFDB-Daten
L_values_lmfdb = freese_formula(lmfdb_zeros, A, beta, a, b)

# Visualisierung der Kohärenzlängen für beide Datensätze
plt.figure(figsize=(12, 6))

# Plot für Odlyzko-Daten
plt.subplot(1, 2, 1)
plt.plot(zeros_odlyzko, L_values_odlyzko, label="Freese-Formel - Odlyzko")
plt.xlabel("Nullstellen (Reelle Teile) Odlyzko")
plt.ylabel("Kohärenzlänge L(N)")
plt.title("Vergleich von Kohärenzlängen (Odlyzko)")
plt.legend()
plt.grid(True)

# Plot für LMFDB-Daten
plt.subplot(1, 2, 2)
plt.plot(lmfdb_zeros, L_values_lmfdb, label="Freese-Formel - LMFDB", color='orange')
plt.xlabel("Nullstellen (Reelle Teile) LMFDB")
plt.ylabel("Kohärenzlänge L(N)")
plt.title("Vergleich von Kohärenzlängen (LMFDB)")
plt.legend()
plt.grid(True)

# Anzeige des Plots
plt.tight_layout()
plt.show()

# Ausgabe der berechneten Kohärenzlängen für die ersten 10 Nullstellen beider Datensätze
print("Die ersten 10 Kohärenzlängen für Odlyzko Nullstellen:")
print(L_values_odlyzko[:10])

print("Die ersten 10 Kohärenzlängen für LMFDB Nullstellen:")
print(L_values_lmfdb[:10])
```
