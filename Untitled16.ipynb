```python
import numpy as np
import matplotlib.pyplot as plt
import multiprocessing as mp
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# üìÇ **Lade die 10 Mio Nullstellen aus Rainbowtable**
file_path = "/content/drive/MyDrive/rainbowtable_10mio.txt"
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üìå **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * np.power(n, -0.5) + B * np.power(n, -1) + C

# üìå **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * np.power(n, -0.5) + B * np.power(n, -1) + C + w * np.cos(n * w + phi)

# üìå **Alternative Form: Zus√§tzliche Potenz statt Oszillation**
def freese_alternative(n, A, B, C, D):
    return A * np.power(n, -0.5) + B * np.power(n, -1) + C + D * np.power(n, -1.5)

# üîç **Parallelisierte Parameter-Optimierung**
def optimize_function(func, p0, bounds=None):
    """ Optimiert eine gegebene Funktion mit Curve Fit in einem separaten Prozess """
    with mp.Pool(processes=1) as pool:
        result = pool.apply(curve_fit, args=(func, n_values, nullstellen_abstaende, p0, bounds))
    return result

print("üîç Starte Optimierung f√ºr FFS...")
params_ffs, _ = optimize_function(freese_standard, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

print("üîç Starte Optimierung f√ºr FFO...")
params_ffo, _ = optimize_function(freese_oscillating, p0=[A_ffs, B_ffs, C_ffs, 0.01, 0])
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, œÜ={phi_ffo:.6f}")

print("üîç Starte Optimierung f√ºr alternative Form...")
params_alt, _ = optimize_function(freese_alternative, p0=[A_ffs, B_ffs, C_ffs, 0])
A_alt, B_alt, C_alt, D_alt = params_alt
print(f"‚úÖ Optimierte Alternative: A={A_alt:.6f}, B={B_alt:.6f}, C={C_alt:.6f}, D={D_alt:.6f}")

# üîç **Fehlermetriken berechnen**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
error_alt = np.mean(np.abs(nullstellen_abstaende - freese_alternative(n_values, *params_alt)))

print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")
print(f"‚ö° Alternative Fehler: {error_alt:.6f}")

# üìä **Fehlerspektrum mit Fourier-Analyse**
errors_ffs = nullstellen_abstaende - freese_standard(n_values, *params_ffs)
frequenzen = rfftfreq(len(errors_ffs))
spektrum = np.abs(rfft(errors_ffs))

# üìà **Histogramm der Fehlerverteilung**
plt.figure(figsize=(10,5))
plt.hist(errors_ffs, bins=100, color="blue", alpha=0.7, label="FFS Fehler")
plt.xlabel("Fehler")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm der Fehlerverteilung (FFS)")
plt.legend()
plt.show()

# üìâ **Fourier-Spektrum der Fehlerfunktion**
plt.figure(figsize=(10,5))
plt.plot(frequenzen[:1000], spektrum[:1000], color="red")
plt.yscale("log")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Fehler (FFS)")
plt.show()

# üìä **Vergleich der Modelle mit echten Daten**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:5000], nullstellen_abstaende[:5000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:5000], freese_standard(n_values[:5000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:5000], freese_oscillating(n_values[:5000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.plot(n_values[:5000], freese_alternative(n_values[:5000], *params_alt), label="Alternative", color="green", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabst√§nde vs. FFS, FFO & Alternative")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# üìÇ Dateipfade
file_path_zeros = "/content/drive/MyDrive/rainbowtable_10mio.txt"
file_path_primes = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen1mio.txt"

# üìÇ Lade Daten
print(f"üìÇ Lade Nullstellen aus {file_path_zeros} ...")
nullstellen = np.loadtxt(file_path_zeros)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)

print(f"üìÇ Lade Primzahlen aus {file_path_primes} ...")
primzahlen = np.loadtxt(file_path_primes)

# üìä **Histogramm der Nullstellen-Abst√§nde**
plt.figure(figsize=(10, 5))
plt.hist(nullstellen_abstaende, bins=100, color='blue', alpha=0.7, label="Nullstellen-Abst√§nde")
plt.xlabel("Abstand")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm der Nullstellen-Abst√§nde (10Mio Rainbowtable)")
plt.legend()
plt.show()

# üìà **3D-Doppelhelix der Riemann-Nullstellen**
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 10*np.pi, len(nullstellen[:1000]))
x = np.cos(t) * nullstellen[:1000]
y = np.sin(t) * nullstellen[:1000]
z = nullstellen[:1000]
ax.plot_trisurf(x, y, z, cmap=cm.viridis, edgecolor='none')
ax.set_xlabel("X-Koordinate (Re Nullstellen)")
ax.set_ylabel("Y-Koordinate (Im Nullstellen)")
ax.set_zlabel("Z-Koordinate (Index)")
ax.set_title("3D-Visualisierung der Riemann-Nullstellen als Doppelhelix")
plt.show()

# üì° **Fermat-Spirale mit Primzahlen**
theta = np.sqrt(primzahlen[:5000]) * np.pi
r = np.sqrt(primzahlen[:5000])
x = r * np.cos(theta)
y = r * np.sin(theta)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, primzahlen[:5000], c=primzahlen[:5000], cmap='plasma')
ax.set_xlabel("X-Koordinate (Re Primzahlen)")
ax.set_ylabel("Y-Koordinate (Im Primzahlen)")
ax.set_zlabel("Z-Koordinate (Primzahl)")
ax.set_title("Primzahlen & Zeta-Nullstellen auf der Fermat-Spirale")
plt.show()

# üî• **Mandelbrotmenge mit Nullstellen-Overlay**
def mandelbrot(c, max_iter=500):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z**2 + c
    return max_iter

resolution = 500
x_min, x_max = -2, 1
y_min, y_max = -1.5, 1.5
x, y = np.linspace(x_min, x_max, resolution), np.linspace(y_min, y_max, resolution)
X, Y = np.meshgrid(x, y)
C = X + 1j * Y
mandelbrot_set = np.array([[mandelbrot(c) for c in row] for row in C])

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(mandelbrot_set, extent=(x_min, x_max, y_min, y_max), cmap='inferno', origin='lower')
ax.set_xlabel("Re(c)")
ax.set_ylabel("Im(c)")
ax.set_title("Mandelbrotmenge & Riemannsche Nullstellen")

# Overlay der Nullstellen
nullstellen_overlay = nullstellen[:100] / max(nullstellen[:100]) * (x_max - x_min) + x_min
ax.scatter(nullstellen_overlay, np.zeros_like(nullstellen_overlay), color='cyan', label="Riemann-Nullstellen", s=10)
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Daten f√ºr die Nullstellen simulieren
n_values = np.arange(1, 5000)
real_parts = 0.5 * np.ones_like(n_values)
imaginary_parts = np.linspace(0, 5000, len(n_values))

# 3D-Plot der Nullstellen in der komplexen Ebene
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(real_parts, imaginary_parts, zs=0, zdir='z', label='Nullstellen')

ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_zlabel('Index')
ax.set_title("3D-Darstellung der Riemann-Nullstellen")
plt.show()

# Mandelbrotmenge mit Nullstellen als Overlay
def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z**2 + c
    return max_iter

# Bildgr√∂√üe
width, height = 600, 600
xmin, xmax = -2, 1
ymin, ymax = -1.5, 1.5
max_iter = 500

# Mandelbrotmenge berechnen
image = np.zeros((height, width))
for x in range(width):
    for y in range(height):
        cx = xmin + (x / width) * (xmax - xmin)
        cy = ymin + (y / height) * (ymax - ymin)
        color = mandelbrot(complex(cx, cy), max_iter)
        image[y, x] = color

# Nullstellen auf die Mandelbrotmenge plotten
plt.figure(figsize=(8, 8))
plt.imshow(image, cmap='hot', extent=[xmin, xmax, ymin, ymax])
plt.colorbar(label="Iterationen bis Divergenz")

# Nullstellen hinzuf√ºgen
plt.scatter(real_parts - 1, imaginary_parts / 5000, color='cyan', s=5, label="Riemann-Nullstellen")
plt.legend()
plt.title("Mandelbrotmenge & Riemannsche Nullstellen")
plt.xlabel("Re(c)")
plt.ylabel("Im(c)")
plt.show()
```
