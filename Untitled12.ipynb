```python
# ✅ 1. Importieren der benötigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp, expon, norm
from tqdm import tqdm
from google.colab import drive

# ✅ 2. Google Drive mounten, um auf die Datei zuzugreifen
drive.mount('/content/drive')

# ✅ 3. Nullstellen-Daten aus der Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"  # Pfad zur Datei
nullstellen = np.loadtxt(file_path)  # Laden der Daten

# ✅ 4. Berechnung der Nullstellenabstände (erste Differenzen)
abstaende = np.diff(nullstellen)

# ✅ 5. Erste 10 Abstände ausgeben
print("Erste 10 Nullstellenabstände:", abstaende[:10])

# ✅ 6. Normalisierung der Abstände
abstaende_norm = abstaende / np.mean(abstaende)

# ✅ 7. Multifraktale Analyse mit Box-Counting-Methode
def multifraktale_analyse(data, q_values, segment_sizes):
    D_q = []
    for q in tqdm(q_values, desc="Multifraktale Dimension berechnen"):
        D_q_vals = []
        for seg in segment_sizes:
            partitioned_data = np.array_split(data, len(data) // seg)
            mu = np.array([np.sum(part**q) for part in partitioned_data if len(part) > 0])
            mu = mu / np.sum(mu)
            S_q = np.sum(mu**q)
            D_q_vals.append(np.log(S_q) / np.log(1 / seg) if S_q > 0 else 0)
        D_q.append(np.mean(D_q_vals))
    return D_q, q_values

# ✅ 8. Definition der Skalenparameter
q_values = np.linspace(-3, 3, 10)  # Weniger q-Werte für schnellere Berechnung
segment_sizes = np.linspace(50, 5000, 10).astype(int)  # Weniger Segmentgrößen für schnelleres Ergebnis

# ✅ 9. Berechnung der multifraktalen Dimensionen
D_q, q_values = multifraktale_analyse(abstaende_norm, q_values, segment_sizes)

# ✅ 10. Visualisierung des Multifraktal-Spektrums
plt.figure(figsize=(10, 6))
plt.plot(q_values, D_q, 'bo-', label="Multifraktale Dimension D(q)")
plt.xlabel("q-Werte")
plt.ylabel("Multifraktale Dimension D(q)")
plt.title("Multifraktale Analyse der Zeta-Nullstellenabstände")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ✅ 1. Importieren der Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from google.colab import drive

# ✅ 2. Google Drive mounten (falls Daten aus Drive geladen werden müssen)
drive.mount('/content/drive')

# ✅ 3. Nullstellen-Daten aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"  # Pfad zur Datei
nullstellen = np.loadtxt(file_path)  # Laden der Daten
abstaende = np.diff(nullstellen)  # Nullstellenabstände berechnen

# ✅ 4. Indizes für die Nullstellen definieren
N_values = np.arange(1, len(abstaende) + 1)

# ✅ 5. Modell für die Doppelhelix definieren
def double_helix(N, A, beta, f, phi, lambda_z):
    R = A * N**beta  # Skalierender Radius
    x = R * np.cos(2 * np.pi * f * N + phi)
    y = R * np.sin(2 * np.pi * f * N + phi)
    z = lambda_z * N  # Vertikale Skalierung
    return x, y, z

# ✅ 6. Parameter-Anpassung mit Curve-Fit
def fit_function(N, A, beta, f, phi, lambda_z):
    x, y, z = double_helix(N, A, beta, f, phi, lambda_z)
    return np.concatenate((x, y, z))

# ✅ 7. Startparameter
initial_params = [1.0, -0.1, 0.01, 0.0, 0.001]

# ✅ 8. Curve Fit durchführen
params, _ = curve_fit(fit_function, N_values, np.concatenate((abstaende, abstaende, abstaende)), p0=initial_params)

# ✅ 9. Optimierte Parameter ausgeben
A_opt, beta_opt, f_opt, phi_opt, lambda_opt = params
print("\n📊 **Optimierte Parameter für die Doppelhelix:**")
print(f"A = {A_opt:.6f}")
print(f"Beta = {beta_opt:.6f}")
print(f"Frequenz f = {f_opt:.6f}")
print(f"Phase phi = {phi_opt:.6f}")
print(f"Lambda (Höhenwachstum) = {lambda_opt:.6f}")

# ✅ 10. Visualisierung der Doppelhelix
x_fit, y_fit, z_fit = double_helix(N_values, A_opt, beta_opt, f_opt, phi_opt, lambda_opt)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_fit, y_fit, z_fit, 'b-', label="Angepasste Doppelhelix")
ax.set_xlabel("X-Koordinate")
ax.set_ylabel("Y-Koordinate")
ax.set_zlabel("Z-Koordinate (Nullstellen-Index)")
ax.set_title("📈 Optimiertes Modell der Zeta-Doppelhelix")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.special import jn

# ✅ Datei mit echten Nullstellen laden
file_path = "/content/drive/MyDrive/zeros6.txt"
data = np.loadtxt(file_path)
N_values = np.arange(1, len(data) + 1)  # Indexwerte für Nullstellen
abstaende = np.diff(data)  # Nullstellen-Abstände berechnen

# ✅ Definition der Fibonacci-Freese-Formel mit Korrekturtermen
def fibonacci_freese_korrigiert(N, A, f, C, D, B, omega, phi):
    return A * N**f + C * np.log1p(N) + D / N + B * jn(0, omega * N + phi)

# ✅ Curve-Fit mit Startwerten
initial_params = [-0.1, 0.48, 0.1, 10, 0.01, 0.01, 0]
params, _ = curve_fit(fibonacci_freese_korrigiert, N_values[1:], abstaende, p0=initial_params, maxfev=10000)

# ✅ Extrahierte Parameter ausgeben
A, f, C, D, B, omega, phi = params
print(f"📊 Optimierte Parameter der Freese-Formel:")
print(f"🔹 A = {A:.6f}")
print(f"🔹 Experimenteller Exponent f = {f:.6f} (Erwartet: 0.484964)")
print(f"🔹 Logarithmischer Term C = {C:.6f}")
print(f"🔹 Korrekturterm D/N = {D:.6f}")
print(f"🔹 Sinus-Amplitude B = {B:.6f}")
print(f"🔹 Sinus-Frequenz omega = {omega:.6f}")
print(f"🔹 Sinus-Phase phi = {phi:.6f}")

# ✅ Vergleich mit den realen Nullstellen-Abständen
fitted_values = fibonacci_freese_korrigiert(N_values[1:], A, f, C, D, B, omega, phi)

# ✅ Plot der realen Nullstellen-Abstände vs. Modell
plt.figure(figsize=(10, 5))
plt.plot(N_values[1:], abstaende, label="Reale Nullstellen-Abstände", alpha=0.6)
plt.plot(N_values[1:], fitted_values, label="Fibonacci-Freese-Modell", linestyle="dashed")
plt.xlabel("Index der Nullstelle (N)")
plt.ylabel("Kohärenzlänge L(N)")
plt.title("Vergleich: Reale Nullstellen vs. Fibonacci-Freese-Formel")
plt.legend()
plt.show()

# ✅ Abweichung berechnen
abweichung = np.abs(abstaende - fitted_values) / abstaende
print(f"⚠️ Mittlere Abweichung: {np.mean(abweichung):.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# 📂 Dateipfad zur Nullstellen-Datei (Google Drive-Integration nötig)
file_path = "/content/drive/MyDrive/zeros6.txt"

# 📂 **Lade Nullstellen & berechne Abstände**
print(f"📂 Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"✅ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# 📌 **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * np.power(n, -1/2, where=n!=0) + B * np.power(n, -1, where=n!=0) + C

# 📌 **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * np.power(n, -1/2, where=n!=0) + B * np.power(n, -1, where=n!=0) + C + w * np.cos(n * w + phi)

# 🔍 **Fourier-Analyse zur Frequenzbestimmung für FFO**
frequenzen = rfftfreq(len(nullstellen_abstaende))
spektrum = np.abs(rfft(nullstellen_abstaende))
dominante_freq_index = np.argmax(spektrum[1:]) + 1
w_auto = frequenzen[dominante_freq_index]
print(f"✅ Automatisch ermittelte Frequenz für FFO: w = {w_auto:.6f}")

# 🔍 **Optimierung der Parameter für FFS**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"✅ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# 🔍 **Optimierung der Parameter für FFO**
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=[2, 0.9, 1, w_auto, 0])
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"✅ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, φ={phi_ffo:.6f}")

# 🔍 **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
print(f"⚡ FFS Fehler: {error_ffs:.6f}")
print(f"⚡ FFO Fehler: {error_ffo:.6f}")

# 📊 **Visualisierung**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abstände", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabstände vs. FFS & FFO")
plt.legend()
plt.show()
```
