```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq

# üìÇ **Dateipfad f√ºr Nullstellen**
file_path = "/content/drive/MyDrive/zeros6.txt"

# üìÇ **Lade Nullstellen & berechne Abst√§nde**
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üîç **FFT-Analyse zur Frequenzbestimmung**
print("üîç Berechne dominierende Frequenz mit Fourier-Analyse...")
N = len(nullstellen_abstaende)
freqs = fftfreq(N, d=1)
fft_values = np.abs(fft(nullstellen_abstaende))

# **Finde die dominante Frequenz, ohne `0.0` als Ergebnis**
fft_values[0] = 0  # Setze die DC-Komponente auf 0 (Offset-Korrektur)
positive_freqs = freqs[1:N//2]
positive_fft_values = fft_values[1:N//2]

# Falls die FFT keine Frequenz erkennt, setzen wir eine minimale Frequenz
if np.max(positive_fft_values) > 0:
    dominant_freq = positive_freqs[np.argmax(positive_fft_values)]
else:
    dominant_freq = 0.001  # **Fallback-Wert f√ºr `w`**

w_auto = abs(dominant_freq)
print(f"‚úÖ Automatisch ermittelte Frequenz f√ºr FFO: w = {w_auto:.6f}")

# üìå **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * n**(-1/2) + B * n**(-1) + C

# üìå **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * n**(-1/2) + B * n**(-1) + C + w * np.cos(n * w + phi)

# üîç **Parameteroptimierung**
n_values = np.arange(1, len(nullstellen_abstaende) + 1)

# ‚úÖ **Optimierung der FFS-Parameter**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# ‚úÖ **Optimierung der FFO-Parameter (inkl. w aus FFT)**
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=[2, 0.9, 1, w_auto, 0], bounds=([-np.inf, -np.inf, -np.inf, 0.0001, -np.pi], [np.inf, np.inf, np.inf, np.pi, np.pi]))
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, œÜ={phi_ffo:.6f}")

# üîç **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")

# üìä **Visualisierung der Ergebnisse**
plt.figure(figsize=(10, 5))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS-Fit", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO-Fit (optimiert)", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.stats import linregress
import math

# Beispielhafte Nullstellen (k√∂nnte von der LMFDB oder Odlyzko-Datenquelle stammen)
# Hier nehmen wir nur eine kleine Menge an Nullstellen f√ºr das Beispiel
zeros = [
    14.134725142, 21.022039639, 25.010857580, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478
]

# Berechne die Abst√§nde zwischen den Nullstellen
distances = np.diff(zeros)

# Fourier-Transformation der Abst√§nde
N = len(distances)
yf = fft(distances)
xf = fftfreq(N, 1)

# Plot der Fourier-Transformation
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(xf, np.abs(yf), label="Fourier Transform")
plt.title("Fourier-Analyse der Nullstellen-Abst√§nde")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()

# Berechnung der Fibonacci-Skalierung nach der Freese-Formel (L(N) = A * N^Œ≤)
# Sch√§tzung der Fibonacci-Skalierungsparameter mit Regressionsanalyse
N_values = np.arange(1, len(distances) + 1)
slope, intercept, r_value, p_value, std_err = linregress(np.log(N_values), np.log(distances))

# Berechnung von A und Œ≤
A = np.exp(intercept)
beta = slope

# Ausgabe der berechneten Parameter und Bewertung des Zusammenhangs
print(f"Berechneter Skalierungsfaktor A: {A}")
print(f"Berechneter Exponent beta: {beta}")
print(f"Bestimmtheitsma√ü R¬≤: {r_value**2}")
print(f"P-Wert: {p_value}")
print(f"Standardfehler: {std_err}")

# Plotten der Originaldaten und der Modellvorhersage (Freese-Formel)
model_prediction = A * N_values**beta
plt.subplot(1, 2, 2)
plt.plot(N_values, distances, 'bo', label="Originaldaten")
plt.plot(N_values, model_prediction, 'r-', label="Modellvorhersage (Freese-Formel)")
plt.title("Vergleich von Originaldaten und Modellvorhersage")
plt.xlabel("Anzahl der Nullstellen (N)")
plt.ylabel("Abstand der Nullstellen (L(N))")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Ausgabe der Ergebnisse
print(f"\nZusammenfassung der Analyse:")
print(f"Der Bestimmtheitsma√ü R¬≤ betr√§gt {r_value**2:.4f}, was auf eine starke Korrelation zwischen den Abst√§nden der Nullstellen und der Fibonacci-Skalierung hinweist.")
print(f"Der P-Wert von {p_value:.4f} legt nahe, dass das Modell signifikant ist.")
print(f"Die Standardabweichung des Fehlers betr√§gt {std_err:.4f}, was die Pr√§zision der Sch√§tzungen zeigt.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# üìÇ Dateipfad zur Nullstellen-Datei (Google Drive-Integration n√∂tig)
file_path = "/content/drive/MyDrive/zeros6.txt"

# üìÇ **Lade Nullstellen & berechne Abst√§nde**
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üìå **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * np.power(n, -1/2, where=n!=0) + B * np.power(n, -1, where=n!=0) + C

# üìå **FFO (Freese-Formel mit mehreren Oszillationen)**
def freese_oscillating(n, A, B, C, *freq_params):
    result = A * np.power(n, -1/2, where=n!=0) + B * np.power(n, -1, where=n!=0) + C
    for i in range(0, len(freq_params), 2):
        w, phi = freq_params[i], freq_params[i+1]
        result += w * np.cos(n * w + phi)
    return result

# üîç **Erweiterte Fourier-Analyse**
frequenzen = rfftfreq(len(nullstellen_abstaende))
spektrum = np.abs(rfft(nullstellen_abstaende))

# W√§hle die 3 dominantesten Frequenzen
top_indices = np.argsort(spektrum[1:])[-3:] + 1
dominante_frequenzen = frequenzen[top_indices]
print(f"‚úÖ Dominante Frequenzen f√ºr FFO: {dominante_frequenzen}")

# üîç **Optimierung der Parameter f√ºr FFS**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# üîç **Optimierung der Parameter f√ºr FFO mit mehreren Oszillationen**
initial_params_ffo = [A_ffs, B_ffs, C_ffs] + [0.1, 0] * len(dominante_frequenzen)
bounds_ffo = ([-np.inf, -np.inf, -np.inf] + [0, -np.pi] * len(dominante_frequenzen),
              [np.inf, np.inf, np.inf] + [np.pi, np.pi] * len(dominante_frequenzen))

params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=initial_params_ffo, bounds=bounds_ffo)
A_ffo, B_ffo, C_ffo = params_ffo[:3]
freq_params_ffo = params_ffo[3:]
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}")
for i in range(0, len(freq_params_ffo), 2):
    print(f"‚úÖ w_{i//2+1} = {freq_params_ffo[i]:.6f}, œÜ_{i//2+1} = {freq_params_ffo[i+1]:.6f}")

# üîç **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")

# üìä **Visualisierung**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabst√§nde vs. FFS & FFO mit mehreren Frequenzen")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq
from multiprocessing import Pool

# üìÇ Lade Nullstellen aus Google Drive
file_path = "/content/drive/MyDrive/zeros6.txt"
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üìå **Freese-Formeln (Fix f√ºr Nullwerte)**
def freese_standard(n, A, B, C):
    n_safe = np.where(n == 0, 1e-10, n).astype(float)  # Schutz gegen Nullwerte
    return A * np.power(n_safe, -1/2) + B * np.power(n_safe, -1) + C

def freese_oscillating(n, A, B, C, w, phi):
    n_safe = np.where(n == 0, 1e-10, n).astype(float)
    return A * np.power(n_safe, -1/2) + B * np.power(n_safe, -1) + C + w * np.cos(n_safe * w + phi)

# üîç **Verbesserte Fourier-Analyse f√ºr w**
frequenzen = rfftfreq(len(nullstellen_abstaende))
spektrum = np.abs(rfft(nullstellen_abstaende))
dominante_freq_index = np.argmax(spektrum[1:]) + 1
w_auto = frequenzen[dominante_freq_index]
print(f"‚úÖ Automatisch ermittelte Frequenz f√ºr FFO: w = {w_auto:.6f}")

# üîç **Optimierung der FFS-Parameter**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# üîç **Grid-Search f√ºr FFO (mehrere w- & œÜ-Werte testen)**
w_values = np.linspace(w_auto * 0.8, w_auto * 1.2, 10)  # Bereich rund um w_auto
phi_values = np.linspace(-np.pi, np.pi, 10)  # 10 Werte f√ºr œÜ

def optimize_ffo(params):
    A, B, C, w, phi = params
    error = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, A, B, C, w, phi)))
    return (w, phi, error)

param_grid = [(A_ffs, B_ffs, C_ffs, w, phi) for w in w_values for phi in phi_values]

# **Parallelisierte Berechnung mit Multiprocessing**
with Pool(processes=4) as pool:
    results = pool.map(optimize_ffo, param_grid)

# üîç **Bestes Ergebnis ausw√§hlen**
best_w, best_phi, best_error = min(results, key=lambda x: x[2])
print(f"‚úÖ Bestes w: {best_w:.6f}, Bestes œÜ: {best_phi:.6f}, Fehler: {best_error:.6f}")

# üîç **Optimierte FFO-Parameter**
params_ffo = [A_ffs, B_ffs, C_ffs, best_w, best_phi]
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}, w={best_w:.6f}, œÜ={best_phi:.6f}")

# üîç **Fehlermetriken**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = best_error
print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")

# üìä **Visualisierung**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:1000], nullstellen_abstaende[:1000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:1000], freese_standard(n_values[:1000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:1000], freese_oscillating(n_values[:1000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabst√§nde vs. Automatisierte FFS & FFO")
plt.legend()
plt.show()
```

```python
# üìÇ Google Drive mounten
from google.colab import drive
drive.mount('/content/drive')

import numpy as np
import requests
import re
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# üìÇ **Dateipfade f√ºr die Nullstellen**
odlyzko_file = "/content/drive/MyDrive/zeros6.txt"  # Lokale Datei
lmfdb_url = "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"  # LMFDB-Nullstellen

# üìÇ **1Ô∏è‚É£ LMFDB Nullstellen abrufen & bereinigen**
def load_lmfdb_zeros(url):
    print(f"üì° Lade Nullstellen aus LMFDB: {url}")
    response = requests.get(url)
    if response.status_code != 200:
        print("‚ùå Fehler beim Abrufen der LMFDB-Daten!")
        return []

    # üîç Suche nach Zahlen (entfernt Zeilennummern & HTML-Formatierungen)
    raw_data = re.findall(r"[\d]+\.\d+", response.text)
    lmfdb_zeros = np.array([float(x) for x in raw_data])
    print(f"‚úÖ {len(lmfdb_zeros)} Nullstellen aus LMFDB erfolgreich geladen!")
    return lmfdb_zeros

# üìÇ **2Ô∏è‚É£ Odlyzko Nullstellen laden**
def load_odlyzko_zeros(file_path):
    print(f"üìÇ Lade Nullstellen aus {file_path} ...")
    odlyzko_zeros = np.loadtxt(file_path)
    print(f"‚úÖ {len(odlyzko_zeros)} Nullstellen erfolgreich geladen!")
    return odlyzko_zeros

# üìä **3Ô∏è‚É£ Skalierungscheck & Transformation**
def scale_and_combine(zeros1, zeros2, target_size=10_000_000):
    print("üîç Pr√ºfe Skalierung & kombiniere Datens√§tze...")

    # Pr√ºfe, ob die ersten Nullstellen √ºbereinstimmen
    offset = zeros2[0] - zeros1[-1]
    if abs(offset) > 1e-3:
        print(f"‚ö† Warnung: Skalierung n√∂tig! Offset: {offset}")
        zeros2 -= offset  # Falls n√∂tig, Verschiebung korrigieren

    # Kombiniere beide Datens√§tze
    combined = np.concatenate((zeros1, zeros2))
    combined = np.sort(combined)  # Sortiere Nullstellen

    # Falls zu wenige Datenpunkte, Interpolation
    if len(combined) < target_size:
        print("üìà Interpolation, um 10 Mio Werte zu erreichen...")
        x = np.linspace(0, len(combined)-1, len(combined))
        f_interp = interp1d(x, combined, kind='cubic')
        new_x = np.linspace(0, len(combined)-1, target_size)
        combined = f_interp(new_x)

    print(f"‚úÖ Rainbowtable mit {len(combined)} Nullstellen erstellt!")
    return combined

# üìÇ **üöÄ Hauptprogramm**
lmfdb_zeros = load_lmfdb_zeros(lmfdb_url)
odlyzko_zeros = load_odlyzko_zeros(odlyzko_file)
rainbowtable = scale_and_combine(lmfdb_zeros, odlyzko_zeros, target_size=10_000_000)

# üíæ **4Ô∏è‚É£ Speichern der neuen 10Mio-Rainbowtable**
output_file = "/content/drive/MyDrive/rainbowtable_10mio.txt"
np.savetxt(output_file, rainbowtable)
print(f"‚úÖ Rainbowtable gespeichert unter: {output_file}")

# üìä **5Ô∏è‚É£ Visualisierung**
plt.figure(figsize=(10,5))
plt.hist(np.diff(rainbowtable), bins=100, alpha=0.7, color="blue", label="Nullstellen-Abst√§nde")
plt.xlabel("Abstand")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm der Nullstellen-Abst√§nde (10Mio Rainbowtable)")
plt.legend()
plt.show()
```

```python
# üìå Notwendige Bibliotheken laden
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# üìÇ Dateipfad zur Nullstellen-Rainbowtable (Google Drive-Integration n√∂tig)
file_path = "/content/drive/MyDrive/rainbowtable_10mio.txt"

# üìÇ **Lade Nullstellen & berechne Abst√§nde**
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üìå **FFS (Freese-Formel Standard) mit Fix**
def freese_standard(n, A, B, C):
    n = np.asarray(n, dtype=float)  # Sicherstellen, dass n als Float behandelt wird
    return A * n**(-1/2) + B * n**(-1) + C

# üìå **FFO (Freese-Formel mit Oszillation) mit Fix**
def freese_oscillating(n, A, B, C, w, phi):
    n = np.asarray(n, dtype=float)  # Sicherstellen, dass n als Float behandelt wird
    return A * n**(-1/2) + B * n**(-1) + C + w * np.cos(n * w + phi)

# üîç **Fourier-Analyse zur Frequenzbestimmung f√ºr FFO**
def fourier_analyse(abstaende):
    frequenzen = rfftfreq(len(abstaende))
    spektrum = np.abs(rfft(abstaende))
    dominante_freq_index = np.argmax(spektrum[1:]) + 1
    return frequenzen[dominante_freq_index]

# üîç **Bestimme die dominierende Frequenz f√ºr FFO**
w_auto = fourier_analyse(nullstellen_abstaende)
print(f"‚úÖ Automatisch ermittelte Frequenz f√ºr FFO: w = {w_auto:.6f}")

# üîç **Optimierung der Parameter f√ºr FFS**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[10, -10, 0.5])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# üîç **Optimierung der Parameter f√ºr FFO**
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=[10, -10, 0.5, w_auto, 0])
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, œÜ={phi_ffo:.6f}")

# üîç **Fehlermetriken berechnen**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")

# üìä **Visualisierung der Ergebnisse**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:5000], nullstellen_abstaende[:5000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:5000], freese_standard(n_values[:5000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:5000], freese_oscillating(n_values[:5000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabst√§nde vs. FFS & FFO (10Mio Rainbowtable)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import rfft, rfftfreq

# üìÇ **Lade die 10 Mio Nullstellen aus Rainbowtable**
file_path = "/content/drive/MyDrive/rainbowtable_10mio.txt"
print(f"üìÇ Lade Nullstellen aus {file_path} ...")
nullstellen = np.loadtxt(file_path)
nullstellen_abstaende = np.diff(nullstellen)
n_values = np.arange(1, len(nullstellen_abstaende) + 1)
print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")

# üìå **FFS (Freese-Formel Standard)**
def freese_standard(n, A, B, C):
    return A * n**(-1/2) + B * n**(-1) + C

# üìå **FFO (Freese-Formel mit Oszillation)**
def freese_oscillating(n, A, B, C, w, phi):
    return A * n**(-1/2) + B * n**(-1) + C + w * np.cos(n * w + phi)

# üìå **Alternative Form: Zus√§tzliche Potenz statt Oszillation**
def freese_alternative(n, A, B, C, D):
    return A * n**(-1/2) + B * n**(-1) + C + D * n**(-3/2)

# üîç **Optimierung der Parameter f√ºr FFS**
params_ffs, _ = curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[2, 0.9, 1])
A_ffs, B_ffs, C_ffs = params_ffs
print(f"‚úÖ Optimierte FFS-Parameter: A={A_ffs:.6f}, B={B_ffs:.6f}, C={C_ffs:.6f}")

# üîç **Optimierung der Parameter f√ºr FFO**
params_ffo, _ = curve_fit(freese_oscillating, n_values, nullstellen_abstaende, p0=[A_ffs, B_ffs, C_ffs, 0.01, 0])
A_ffo, B_ffo, C_ffo, w_ffo, phi_ffo = params_ffo
print(f"‚úÖ Optimierte FFO-Parameter: A={A_ffo:.6f}, B={B_ffo:.6f}, C={C_ffo:.6f}, w={w_ffo:.6f}, œÜ={phi_ffo:.6f}")

# üîç **Optimierung der alternativen Korrektur**
params_alt, _ = curve_fit(freese_alternative, n_values, nullstellen_abstaende, p0=[A_ffs, B_ffs, C_ffs, 0])
A_alt, B_alt, C_alt, D_alt = params_alt
print(f"‚úÖ Optimierte Alternative: A={A_alt:.6f}, B={B_alt:.6f}, C={C_alt:.6f}, D={D_alt:.6f}")

# üîç **Fehlermetriken berechnen**
error_ffs = np.mean(np.abs(nullstellen_abstaende - freese_standard(n_values, *params_ffs)))
error_ffo = np.mean(np.abs(nullstellen_abstaende - freese_oscillating(n_values, *params_ffo)))
error_alt = np.mean(np.abs(nullstellen_abstaende - freese_alternative(n_values, *params_alt)))

print(f"‚ö° FFS Fehler: {error_ffs:.6f}")
print(f"‚ö° FFO Fehler: {error_ffo:.6f}")
print(f"‚ö° Alternative Fehler: {error_alt:.6f}")

# üìä **Fehlerspektrum mit Fourier-Analyse**
errors_ffs = nullstellen_abstaende - freese_standard(n_values, *params_ffs)
frequenzen = rfftfreq(len(errors_ffs))
spektrum = np.abs(rfft(errors_ffs))

# üìà **Histogramm der Fehlerverteilung**
plt.figure(figsize=(10,5))
plt.hist(errors_ffs, bins=100, color="blue", alpha=0.7, label="FFS Fehler")
plt.xlabel("Fehler")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm der Fehlerverteilung (FFS)")
plt.legend()
plt.show()

# üìâ **Fourier-Spektrum der Fehlerfunktion**
plt.figure(figsize=(10,5))
plt.plot(frequenzen[:1000], spektrum[:1000], color="red")
plt.yscale("log")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Fehler (FFS)")
plt.show()

# üìä **Vergleich der Funktionen mit echten Daten**
plt.figure(figsize=(12, 6))
plt.plot(n_values[:5000], nullstellen_abstaende[:5000], label="Echte Abst√§nde", color="black", alpha=0.7)
plt.plot(n_values[:5000], freese_standard(n_values[:5000], *params_ffs), label="FFS", color="blue", linestyle="dashed")
plt.plot(n_values[:5000], freese_oscillating(n_values[:5000], *params_ffo), label="FFO", color="red", linestyle="dashed")
plt.plot(n_values[:5000], freese_alternative(n_values[:5000], *params_alt), label="Alternative", color="green", linestyle="dashed")
plt.xlabel("Index n")
plt.ylabel("Abstand zwischen Nullstellen")
plt.title("Vergleich: Echte Nullstellenabst√§nde vs. FFS, FFO & Alternative")
plt.legend()
plt.show()
```
