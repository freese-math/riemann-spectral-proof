```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import requests

# 📌 Gerät wählen (GPU, falls verfügbar)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🔍 Nutze Gerät: {device}")

# 📌 Zeta-Nullstellen von Odlyzko's Datenbank laden
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(URL)
zeta_zeros = torch.tensor([float(line.strip()) for line in response.text.split("\n") if line.strip()], device=device)

# 📌 Anzahl der Nullstellen wählen
N = min(100000, len(zeta_zeros))  # 100K Nullstellen testen
zeta_zeros = zeta_zeros[:N]
print(f"✅ Erfolgreich {N} Zeta-Nullstellen geladen!")

# 📌 Sparse Hamilton-Operator definieren
diagonal = zeta_zeros
off_diagonal = torch.ones(N-1, device=device)
H = torch.diag(diagonal) + torch.diag(off_diagonal, 1) + torch.diag(off_diagonal, -1)

# 📌 Eigenwerte mit torch.linalg.eigvalsh berechnen (Optimierung: direkt für Sparse-Matrix)
print(f"⏳ Berechne {N} Eigenwerte...")
eigenvalues = torch.linalg.eigvalsh(H)
print(f"✅ Erste 10 Eigenwerte: {eigenvalues[:10].cpu().numpy()}")

# 📌 Spurformel berechnen
t_values = torch.linspace(0.01, 10, 100, device=device)
trace_values = torch.tensor([torch.sum(torch.exp(-t * eigenvalues)) for t in t_values], device=device)

# 📌 Beta-Korrektur mit 1/129.4
beta = 1/129.4
corrected_trace = trace_values * torch.exp(beta * torch.pi * 1j)

# 📌 Fehleranalyse: Abweichung zwischen Original und Korrektur
error = torch.abs(trace_values - corrected_trace)

# 📌 Plots
plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), trace_values.cpu().real, 'b-', label="Numerische Spurformel")
plt.plot(t_values.cpu(), corrected_trace.cpu().real, 'r--', label=r"Mit $e^{\beta \pi i}$-Korrektur ($\beta=1/129.4$)")
plt.xlabel("t-Wert")
plt.ylabel("Spurformel-Summe")
plt.title("Vergleich der Spurformel mit und ohne Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), error.cpu(), 'k-', label="Abweichung")
plt.axhline(0, color='r', linestyle='dashed', label="Perfekte Übereinstimmung")
plt.xlabel("t-Wert")
plt.ylabel("Fehler")
plt.title("Fehleranalyse der Beta-Korrektur mit $1/129.4$")
plt.legend()
plt.grid()
plt.show()

# 📌 Maximale und mittlere Abweichung berechnen
max_error = torch.max(error).item()
mean_error = torch.mean(error).item()
print(f"🔍 Maximale Abweichung: {max_error:.8f}")
print(f"🔍 Mittlere Abweichung: {mean_error:.8f}")
```

```python
# 📌 Notwendige Bibliotheken importieren
import torch
import numpy as np
import scipy.optimize as opt
import scipy.fftpack as fft
import matplotlib.pyplot as plt
import requests

# 📌 Gerät wählen (GPU falls verfügbar)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🔍 Nutze Gerät: {device}")

# 📌 Zeta-Nullstellen von Odlyzko's Datenbank laden
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(URL)
zeta_zeros = torch.tensor([float(line.strip()) for line in response.text.split("\n") if line.strip()], device=device)

# 📌 Anzahl der Nullstellen wählen
N = min(30000, len(zeta_zeros))
zeta_zeros = zeta_zeros[:N]
print(f"✅ Erfolgreich {N} Zeta-Nullstellen geladen!")

# 📌 Hamilton-Operator definieren (tridiagonal)
diagonal = zeta_zeros
off_diagonal = torch.ones(N-1, device=device)
H = torch.diag(diagonal) + torch.diag(off_diagonal, 1) + torch.diag(off_diagonal, -1)

# 📌 Eigenwerte berechnen
print(f"⏳ Berechne {N} Eigenwerte...")
eigenvalues = torch.linalg.eigvalsh(H)
print(f"✅ Erste 10 Eigenwerte: {eigenvalues[:10].cpu().numpy()}")

# 📌 Spurformel berechnen
t_values = torch.linspace(0.01, 10, 100, device=device)
trace_values = torch.tensor([torch.sum(torch.exp(-t * eigenvalues)) for t in t_values], device=device)

# 📌 Fit-Funktion: Potenzgesetz
def power_law(t, A, C, gamma):
    return A * t**(-gamma) + C

# 📌 Fit-Funktion: Exponential
def exponential_law(t, A, C, gamma):
    return A * np.exp(-gamma * t) + C

# 📌 Daten für Fit vorbereiten
t_numpy = t_values.cpu().numpy()
trace_numpy = trace_values.cpu().numpy()

# 📌 Curve Fit für Potenzgesetz
popt_power, _ = opt.curve_fit(power_law, t_numpy, trace_numpy, p0=[1.0, 0.1, 1.5])

# 📌 Curve Fit für Exponential
popt_exp, _ = opt.curve_fit(exponential_law, t_numpy, trace_numpy, p0=[1.0, 0.1, 1.5])

# 📌 Fit-Ergebnisse ausgeben
A_power, C_power, gamma_power = popt_power
A_exp, C_exp, gamma_exp = popt_exp
print(f"📊 Fit-Potenzgesetz: A={A_power:.5f}, C={C_power:.5f}, γ={gamma_power:.5f}")
print(f"📊 Fit-Exponential: A={A_exp:.5f}, C={C_exp:.5f}, γ={gamma_exp:.5f}")

# 📌 Plot der Fits
plt.figure(figsize=(10,5))
plt.scatter(t_numpy, trace_numpy, color='red', label="Numerische β(t)-Werte", s=10)
plt.plot(t_numpy, power_law(t_numpy, *popt_power), 'b-', label="Fit: Potenzgesetz")
plt.plot(t_numpy, exponential_law(t_numpy, *popt_exp), 'g--', label="Fit: Exponential")
plt.xlabel("t")
plt.ylabel("β(t)")
plt.title("Vergleich der Fit-Methoden")
plt.legend()
plt.grid()
plt.show()

# 📌 Fourier-Analyse von β(t)
fft_beta = fft.fft(trace_numpy)
freqs = fft.fftfreq(len(trace_numpy), d=(t_numpy[1] - t_numpy[0]))

plt.figure(figsize=(10,5))
plt.plot(freqs[:len(freqs)//2], np.abs(fft_beta)[:len(freqs)//2], 'b-')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von β(t)")
plt.grid()
plt.show()
```

```python
# 📌 Setup für Google Colab
import numpy as np
import torch
import matplotlib.pyplot as plt
from scipy.stats import linregress, chisquare

# 📌 CUDA aktivieren (falls verfügbar)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🔍 Nutze Gerät: {device}")

# 📌 Echte Zeta-Nullstellen & Primzahlen laden
zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt", max_rows=100000)
primes = np.loadtxt("/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt", max_rows=100000)

# 📌 Beta-Werte zur Analyse
betas = [0.007728, 0.007299, 0.045]

# 📌 Speichert R²- und Chi²-Werte für alle Beta-Skalierungen
r_squared_values = {}
chi_squared_values = {}

plt.figure(figsize=(8, 5))

for beta in betas:
    # 📌 Beta-Skalierung auf Nullstellen anwenden
    scaled_zeros = zeta_zeros * (1 + beta)

    # 📌 Lineare Regression (Korrelation mit Primzahlen)
    slope, intercept, r_value, p_value, std_err = linregress(scaled_zeros, primes)
    r_squared = r_value ** 2
    r_squared_values[beta] = r_squared

    # 📌 Chi²-Test zur Überprüfung der Verteilung
    expected = slope * scaled_zeros + intercept
    chi2_stat, p_val = chisquare(primes[:len(expected)], f_exp=expected)
    chi_squared_values[beta] = (chi2_stat, p_val)

    # 📌 Streudiagramm mit Regressionslinie
    plt.scatter(scaled_zeros[:1000], primes[:1000], label=f"β = {beta} (R² = {r_squared:.6f})", alpha=0.6, s=10)
    plt.plot(scaled_zeros[:1000], expected[:1000], linestyle="dashed")

# 📌 Visualisierung
plt.xlabel("Beta-skalierte Zeta-Nullstellen")
plt.ylabel("Primzahlen")
plt.title("Korrelation zwischen Beta-skalierten Zeta-Nullstellen & Primzahlen")
plt.legend()
plt.grid()
plt.show()

# 📌 Ergebnisse anzeigen
print("📊 Ergebnisse des Signifikanztests:")
for beta, r2 in r_squared_values.items():
    chi2, p = chi_squared_values[beta]
    print(f"Beta = {beta:.6f} → R² = {r2:.8f}, Chi² = {chi2:.4f}, p-Wert = {p:.4e}")

# 📌 Beste Beta-Wahl nach R²
optimal_beta = max(r_squared_values, key=r_squared_values.get)
print(f"\n✅ Beste Beta-Wahl: {optimal_beta} mit R² = {r_squared_values[optimal_beta]:.8f}")

# 📌 Beste Beta-Wahl nach Chi²-Test (niedrigstes Chi², höchster p-Wert)
optimal_beta_chi2 = min(chi_squared_values, key=lambda b: chi_squared_values[b][0])
print(f"✅ Beste Beta-Wahl nach Chi²-Test: {optimal_beta_chi2} mit Chi² = {chi_squared_values[optimal_beta_chi2][0]:.4f}, p-Wert = {chi_squared_values[optimal_beta_chi2][1]:.4e}")
```

```python
# ===============================================================
# Jupyter Notebook zur Signifikanzprüfung der Euler-Freese-Identität
# ===============================================================

import numpy as np
import torch
import scipy.stats as stats
import scipy.fftpack as fft
import matplotlib.pyplot as plt

# -------------------------------
# 1. CUDA Setup (Falls verfügbar)
# -------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🔍 Nutze Gerät: {device}")

# -----------------------------------------
# 2. Daten laden: Zeta-Nullstellen & Primzahlen
# -----------------------------------------
zeta_zeros = np.loadtxt('/content/drive/MyDrive/zeros6.txt', max_rows=100000)
primes = np.loadtxt('/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt', max_rows=100000)

zeta_zeros = torch.tensor(zeta_zeros, device=device)
primes = torch.tensor(primes, device=device)
N = len(zeta_zeros)

# ----------------------------------
# 3. Spurformel & Beta-Korrekturen
# ----------------------------------
t_values = torch.linspace(0.01, 10, 100, device=device)
trace_values = torch.tensor([torch.sum(torch.exp(-t * zeta_zeros)) for t in t_values], device=device)

# Beta-Werte zum Testen
betas = torch.tensor([1/137, 1/129.4, 0.007299, 0.007728, 0.045, 1/2, 7/33300], device=device)
errors = {}
r_squared = {}
chi_squared = {}
p_values = {}

for beta in betas:
    corrected_trace = trace_values * torch.exp(-beta * torch.pi)
    residuals = trace_values - corrected_trace

    # R² berechnen
    ss_res = torch.sum(residuals ** 2)
    ss_tot = torch.sum((trace_values - torch.mean(trace_values)) ** 2)
    r_squared[beta.item()] = 1 - (ss_res / ss_tot)

    # Chi²-Test
    chi_squared[beta.item()] = torch.sum((residuals ** 2) / corrected_trace).item()
    p_values[beta.item()] = stats.chi2.sf(chi_squared[beta.item()], df=N-1)

# ----------------------------------
# 4. Ergebnisse ausgeben
# ----------------------------------
print("\n📊 Ergebnisse des Signifikanztests:")
for beta in betas:
    print(f"Beta = {beta.item():.8f} → R² = {r_squared[beta.item()]:.8f}, Chi² = {chi_squared[beta.item()]:.4f}, p-Wert = {p_values[beta.item()]:.4e}")

# ----------------------------------
# 5. Fourier-Analyse der Spurformel
# ----------------------------------
fft_trace = np.abs(fft.fft(trace_values.cpu().numpy()))
freqs = fft.fftfreq(len(trace_values.cpu().numpy()))

plt.figure(figsize=(8, 5))
plt.plot(freqs, fft_trace, label="Fourier-Spektrum der Spurformel")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Analyse der Euler-Freese-Identität")
plt.legend()
plt.grid()
plt.show()
```

```python
# 📌 Colab Setup: GPU aktivieren und benötigte Bibliotheken installieren
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 📌 Zeta-Nullstellen laden (Pfad anpassen!)
file_path = "/content/drive/MyDrive/zeros6.txt"
zeta_zeros = np.loadtxt(file_path)[:100000]  # Nur 100.000 Werte für Effizienz

# 📌 Test-Beta-Werte (kleinste mögliche Werte)
beta_values = [1/333000, 1/1370000, 7/3330000]

# 📌 Ergebnisse speichern
results = []

# 📌 Signifikanztest für jedes Beta
for beta in beta_values:
    transformed = np.exp(1j * beta * np.pi * zeta_zeros)  # Euler-Freese-Korrektur
    real_part = transformed.real.reshape(-1, 1)  # Reale Achse als Feature
    imag_part = transformed.imag  # Imaginäre Achse als Zielwert

    # 📌 Lineare Regression für Signifikanztest
    model = LinearRegression()
    model.fit(real_part, imag_part)
    r_squared = model.score(real_part, imag_part)

    # 📌 Chi²-Test
    chi2_stat = np.sum((imag_part - model.predict(real_part))**2)
    p_value = stats.chi2.sf(chi2_stat, df=len(imag_part)-1)

    results.append((beta, r_squared, chi2_stat, p_value))

# 📌 Ergebnisse anzeigen
for beta, r2, chi2, p in results:
    print(f"Beta = {beta:.10f} → R² = {r2:.8f}, Chi² = {chi2:.4f}, p-Wert = {p:.4e}")

# 📌 Visualisierung der Ergebnisse
betas = [r[0] for r in results]
r2_values = [r[1] for r in results]

plt.figure(figsize=(8,5))
plt.plot(betas, r2_values, marker='o', linestyle='-', color='b', label="R²-Wert")
plt.xscale("log")
plt.xlabel("Beta-Wert (log)")
plt.ylabel("R²-Qualität")
plt.title("Signifikanzanalyse für kleinste Beta-Werte")
plt.legend()
plt.grid()
plt.show()
```
