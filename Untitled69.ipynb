```python
# ========================
# 🔁 Setup: Bibliotheken
# ========================
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.linalg as la

# ========================
# 📂 Pfade zu Deinen Daten
# ========================
path_beta = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
path_zeros = "/content/drive/MyDrive/zeros6.txt"

# ==========================
# 📊 Daten einlesen & prüfen
# ==========================
beta_raw = pd.read_csv(path_beta, header=None, skiprows=1).squeeze().astype(float)
beta = beta_raw[:1000]  # Anzahl der Nullstellen
with open(path_zeros, "r") as f:
    zeta = np.array([float(line.strip()) for line in f.readlines() if line.strip()])[:1000]

print(f"✅ Beta-Skala: {len(beta)} Werte geladen")
print(f"✅ Zeta-Nullstellen: {len(zeta)} Werte geladen")

# ==========================
# ⚙️ Modifizierter Operator: H × exp(iπβ)
# ==========================
N = len(beta)
phase_factor = np.exp(1j * np.pi * beta)
H = np.diag(zeta)  # Diagonaloperator mit Zeta-Nullstellen

# Modifizierter Operator H'
H_mod = H @ np.diag(phase_factor)

# ==========================
# 🔍 Eigenwerte berechnen
# ==========================
eigenvalues, _ = la.eig(H_mod)
real_parts = np.real(eigenvalues)
imag_parts = np.imag(eigenvalues)

# ==========================
# 🔍 Analyse der Struktur
# ==========================
plt.figure(figsize=(7, 6))
plt.scatter(real_parts, imag_parts, s=3, c='purple', alpha=0.5)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.title("Spektrum des Operators $H e^{i \\pi \\beta(n)}$")
plt.xlabel("Re")
plt.ylabel("Im")
plt.grid(True)
plt.show()

# ==========================
# ✅ Validierung gegen Bedingung
# ==========================
residuum = H_mod + np.eye(N)
norm_residual = np.linalg.norm(residuum)
print(f"✅ Norm von $He^{i\\pi \\beta} + I$: {norm_residual:.4e}")
```

```python
# 📌 Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import norm
from numpy import exp, pi, diag, eye

# 📌 Pfade zu den Datendateien
path_beta = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
path_zeta = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt"

# 📌 Anzahl der Werte
anzahl = 20000

# 📌 Beta-Werte laden
beta = pd.read_csv(path_beta, header=None, skiprows=1).squeeze().astype(float).values[:anzahl]

# 📌 Zeta-Nullstellen laden
with open(path_zeta, "r") as f:
    zeta = np.array([float(line.strip()) for line in f if line.strip()])[:anzahl]

# === OPERATOR-KONSTRUKTION ====================================

# 📌 Operator H definieren: Diagonalmatrix mit Zeta-Nullstellen + Nebenachse
H = diag(zeta) + diag(np.ones(anzahl - 1), k=1) + diag(np.ones(anzahl - 1), k=-1)

# 📌 Diagonale Einheitsmatrix (für I)
I = eye(anzahl)

# 📌 Komplexe Phase: e^{iπβ(n)} für jeden Beta-Wert
phasen = exp(1j * pi * beta)

# 📌 Diagonale Matrix mit den Phasen auf der Diagonalen
P = diag(phasen)

# 📌 Operator-Form: He^{iπβ(n)} + I
operator = H @ P + I

# 📌 Norm als Maß für Abweichung von der Nullmatrix
norm_residual = norm(operator)

# === VISUALISIERUNGEN =======================================

# 1. Einheitskreis: e^{iπβ(n)}
punkte = exp(1j * pi * beta)
plt.figure(figsize=(6, 6))
plt.plot(punkte.real, punkte.imag, "o", markersize=1)
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("Punkte: $e^{i\\pi \\beta(n)}$ auf dem Einheitskreis")
plt.grid(True)
plt.axis("equal")
plt.show()

# 2. Vergleich Beta vs. Zeta
plt.figure(figsize=(6, 5))
plt.plot(beta, label="Beta-Skala", color="blue")
plt.plot(zeta, label="Zeta-Nullstellen", color="red")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.title("Vergleich: Beta-Skala vs. Zeta-Nullstellen")
plt.legend()
plt.grid(True)
plt.show()

# 3. Normausgabe
print("✅ Norm von He^(iπβ(n)) + I:", f"{norm_residual:.4e}")
```

```python
# 📌 Bibliotheken laden
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from sympy import primerange

# 📌 Anzahl der Nullstellen definieren
N = 5000

# 📌 Zeta-Nullstellen (imag-Teile) laden oder simulieren
# (Falls Datei vorhanden: hier ersetzen)
zeta_zeros = np.loadtxt("/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv")[:N]  # <-- ersetzen durch echte Datei falls vorhanden

# 📌 Parameter für W(x)
A = 1.0
B = 0.1
C = 0.01

# 📌 Primzahlen vorbereiten
primes = np.array(list(primerange(2, 1000)))

# 📌 Definition der W(x)-Funktion
def W(x):
    if x <= 0:
        return 0.0
    sum_term = np.sum(np.cos(2 * np.pi * np.log(x) / np.log(primes)))
    return A * sum_term + B * np.log(x) + C * x**(-1.5)

# 📌 Potentialvektor erzeugen
x_vals = np.arange(1, N + 1)
V = np.array([W(x) for x in x_vals])

# 📌 Diskretisierter Hamilton-Operator
diag = V
off_diag = -np.ones(N - 1)
H = np.diag(diag) + np.diag(off_diag, k=1) + np.diag(off_diag, k=-1)

# 📌 Eigenwerte berechnen
eigenvalues, _ = la.eigh(H)

# 📌 Selbstadjungiertheit prüfen
is_self_adjoint = np.allclose(H, H.T.conj())
print(f"✅ Selbstadjungiert: {is_self_adjoint}")
print(f"Erste 10 Eigenwerte:\n{np.round(eigenvalues[:10], 4)}")

# 📌 Abweichung zur kritischen Linie (Re(s)=0.5)
max_deviation = np.max(np.abs(eigenvalues.real - 0.5))
print(f"📏 Max. Abweichung zur kritischen Linie: {max_deviation:.5f}")

# 📌 Plot: Eigenwerte in der komplexen Ebene
plt.figure(figsize=(8, 5))
plt.scatter(eigenvalues.real, eigenvalues.imag, s=10, color="blue", label="Operator-Eigenwerte")
plt.axvline(x=0.5, color="red", linestyle="--", label="Re(s) = 1/2 (krit. Linie)")
plt.xlabel("Re(s)")
plt.ylabel("Im(s)")
plt.title(f"Eigenwerte des Operators mit {N} Punkten\nMax. Abweichung: {max_deviation:.2e}")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten laden ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta = pd.read_csv(data_path, header=None).iloc[:, 0].values
n = np.arange(1, len(beta) + 1)

# === 2. Modellfunktion: Logarithmus + Kosinus + Linear-Term ===
def beta_model(n, A, a, b, f, phi, c):
    return A + a / np.log(n) + b * np.cos(2 * np.pi * f * n + phi) + c * n

# === 3. Fit durchführen ===
p0 = [0, 1e5, 1e4, 0.00001, 0, 0]  # Startwerte für A, a, b, f, phi, c
params, _ = curve_fit(beta_model, n, beta, p0=p0, maxfev=10000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params
beta_fit = beta_model(n, *params)
residual = beta - beta_fit

# === 4. Plotten ===
fig, axs = plt.subplots(2, 2, figsize=(16, 8))

# Original vs Modell
axs[0, 0].plot(n, beta, color='gray', label='Reale Beta(n)-Werte')
axs[0, 0].plot(n, beta_fit, color='blue', label='Modellfunktion')
axs[0, 0].set_title('Vergleich: Modellfunktion vs. Beta(n)-Skala')
axs[0, 0].set_xlabel('n')
axs[0, 0].set_ylabel(r'$\beta(n)$')
axs[0, 0].legend()

# Residuum
axs[0, 1].plot(n, residual, color='purple', label='Residuum (Beta - Modell)')
axs[0, 1].axhline(0, linestyle='--', color='gray')
axs[0, 1].set_title('Residuum zwischen realen und modellierten Beta(n)-Werten')
axs[0, 1].set_xlabel('n')
axs[0, 1].set_ylabel('Differenz')
axs[0, 1].legend()

# Logarithmischer Anteil
axs[1, 0].plot(n, a_fit / np.log(n), color='orange', label=r'$\frac{a}{\log n}$')
axs[1, 0].axhline(0, linestyle='--', color='gray')
axs[1, 0].set_title('Logarithmischer Term')
axs[1, 0].set_xlabel('n')
axs[1, 0].set_ylabel('Wert')
axs[1, 0].legend()

# Kosinus-Term
axs[1, 1].plot(n, b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit), color='red', label=r'$b \cdot \cos(2\pi f n + \varphi)$')
axs[1, 1].axhline(0, linestyle='--', color='gray')
axs[1, 1].set_title('Oszillatorischer Term')
axs[1, 1].set_xlabel('n')
axs[1, 1].set_ylabel('Wert')
axs[1, 1].legend()

plt.tight_layout()
plt.show()

# === 5. Fit-Parameter ausgeben ===
print("✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten sicher laden ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"

# Versuche nur numerische Werte zu laden
df = pd.read_csv(data_path, header=None, dtype=str)
df = df.apply(pd.to_numeric, errors='coerce')  # ungültige Werte zu NaN
df = df.dropna()  # NaN-Zeilen entfernen
beta = df.iloc[:, 0].values
n = np.arange(1, len(beta) + 1)

# === 2. Modellfunktion: A + a/log(n) + b*cos(...) + c*n ===
def beta_model(n, A, a, b, f, phi, c):
    return A + a / np.log(n) + b * np.cos(2 * np.pi * f * n + phi) + c * n

# === 3. Fit durchführen ===
p0 = [0, 1e5, 1e4, 0.00001, 0, 0]
params, _ = curve_fit(beta_model, n, beta, p0=p0, maxfev=10000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params
beta_fit = beta_model(n, *params)
residual = beta - beta_fit

# === 4. Visualisierung ===
fig, axs = plt.subplots(2, 2, figsize=(16, 8))

axs[0, 0].plot(n, beta, color='gray', label='Reale Beta(n)-Werte')
axs[0, 0].plot(n, beta_fit, color='blue', label='Modellfunktion')
axs[0, 0].set_title('Vergleich: Modellfunktion vs. Beta(n)-Skala')
axs[0, 0].legend()

axs[0, 1].plot(n, residual, color='purple', label='Residuum')
axs[0, 1].axhline(0, linestyle='--', color='gray')
axs[0, 1].set_title('Residuum (Beta - Modell)')
axs[0, 1].legend()

axs[1, 0].plot(n, a_fit / np.log(n), color='orange', label=r'$\frac{a}{\log n}$')
axs[1, 0].set_title('Log-Term')

axs[1, 1].plot(n, b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit), color='red', label='Oszillation')
axs[1, 1].set_title('Kosinus-Term')

for ax in axs.flat:
    ax.set_xlabel('n')
    ax.set_ylabel(r'$\beta(n)$')
    ax.legend()
    ax.axhline(0, linestyle='--', color='gray')

plt.tight_layout()
plt.show()

# === 5. Parameter anzeigen ===
print("✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# === CSV laden ===
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta = pd.read_csv(beta_path, header=None).squeeze().astype(float).values

# === Zielklassifikation heuristisch erzeugen (Top 5% als "aktiv") ===
threshold_percentile = 95
cutoff = np.percentile(beta, threshold_percentile)
true_labels = (beta >= cutoff).astype(int)

# === Precision-Recall berechnen ===
precision, recall, thresholds = precision_recall_curve(true_labels, beta)
f1_scores = 2 * (precision * recall) / (precision + recall + 1e-8)

# === Bestes F1 finden ===
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
best_f1 = f1_scores[best_idx]

# === Plot ===
plt.figure(figsize=(8, 5))
plt.plot(recall, precision, label="Precision-Recall", color="blue")
plt.scatter(recall[best_idx], precision[best_idx], color="red", label=f"Max F1 = {best_f1:.2f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision vs. Recall mit F1-Maximum")
plt.legend()
plt.grid(True)
plt.show()

print(f"✅ Optimaler Schwellenwert: {best_threshold:.4f}")
print(f"✅ Maximaler F1-Score: {best_f1:.4f}")
print(f"✅ Klassifikations-Cutoff (Top {100 - threshold_percentile}%): {cutoff:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# === Ursprüngliche Beta-Datei laden
beta_path = "/content/drive/MyDrive/beta_scale_operator_cleaned.csv"
beta = pd.read_csv(beta_path, skiprows=1, header=None).squeeze().astype(float).values

# === Mittelwert entfernen → DC-Filter
beta_centered = beta - np.mean(beta)

# === FFT
N = len(beta_centered)
frequencies = fftfreq(N)
spectrum = fft(beta_centered)
spectrum_abs = np.abs(spectrum[:N//2])
frequencies_pos = frequencies[:N//2]

# === Top 10 Frequenzen (ohne Gleichanteil)
top_k = 10
dominant_indices = np.argsort(spectrum_abs)[-top_k:].tolist()
dominant_indices_sorted = sorted(dominant_indices, key=lambda x: -spectrum_abs[x])

# === Ausgabe der dominanten harmonischen Komponenten
for i in dominant_indices_sorted:
    f = frequencies_pos[i]
    A = spectrum_abs[i]
    phi = np.angle(spectrum[i])
    print(f"f = {f:.5f},  A = {A:.4f},  phi = {phi:.2f}")
```
