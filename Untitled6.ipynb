```python
# ✅ Notwendige Bibliotheken installieren (falls nicht vorhanden)
!pip install numpy scipy matplotlib requests pandas

import numpy as np
import scipy.fftpack
import requests
import pandas as pd
import matplotlib.pyplot as plt

# 📥 1️⃣ LMFDB-NULLSTELLEN LADEN (über API)
def fetch_lmfdb_zeros(limit=1000):
    url = f"https://beta.lmfdb.org/api/zeros/riemann_zeta?limit={limit}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return np.array(data['zeros'])
    else:
        raise ValueError("❌ Fehler beim Abrufen der LMFDB-Nullstellen!")

# 📥 2️⃣ ODLYZKO-NULLSTELLEN LADEN
def fetch_odlyzko_zeros():
    url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
    response = requests.get(url)
    if response.status_code == 200:
        lines = response.text.strip().split("\n")
        return np.array([float(line) for line in lines])
    else:
        raise ValueError("❌ Fehler beim Abrufen der Odlyzko-Nullstellen!")

# 📊 3️⃣ Abstandsberechnung der Nullstellen
def compute_differences(zeros):
    return np.diff(np.sort(zeros))

# 🚀 4️⃣ Fourier-Analyse der Abstände
def compute_fft(differences):
    spectrum = np.abs(scipy.fftpack.fft(differences))
    freqs = scipy.fftpack.fftfreq(len(differences))
    return freqs, spectrum

# 📊 5️⃣ Ergebnisse visualisieren
def plot_fft(freqs, spectrum, title):
    plt.figure(figsize=(10, 5))
    plt.plot(freqs, spectrum, color="blue")
    plt.title(f"Fourier-Spektrum der Nullstellen-Abstände ({title})")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()

# 🔄 Daten abrufen & verarbeiten
lmfdb_zeros = fetch_lmfdb_zeros(200000)  # LMFDB-Daten laden (bis zu 200k Nullstellen)
odlyzko_zeros = fetch_odlyzko_zeros()    # Odlyzko-Daten (2M Nullstellen)

# 🔍 Abstände berechnen
lmfdb_diffs = compute_differences(lmfdb_zeros)
odlyzko_diffs = compute_differences(odlyzko_zeros)

# 🔬 Fourier-Analyse
freqs_lmfdb, spectrum_lmfdb = compute_fft(lmfdb_diffs)
freqs_odlyzko, spectrum_odlyzko = compute_fft(odlyzko_diffs)

# 📊 Plots anzeigen
plot_fft(freqs_lmfdb, spectrum_lmfdb, "LMFDB-Nullstellen")
plot_fft(freqs_odlyzko, spectrum_odlyzko, "Odlyzko-Nullstellen")

# 📝 Ergebnisse in einer Tabelle zusammenfassen
df_results = pd.DataFrame({
    "Datenquelle": ["LMFDB", "Odlyzko"],
    "Anzahl Nullstellen": [len(lmfdb_zeros), len(odlyzko_zeros)],
    "Durchschnittlicher Abstand": [np.mean(lmfdb_diffs), np.mean(odlyzko_diffs)]
})

import ace_tools as tools
tools.display_dataframe_to_user(name="Nullstellen-Vergleich", dataframe=df_results)

print("✅ Analyse abgeschlossen!")
```

```python
# Installiere die notwendigen Bibliotheken
!pip install mpmath

import requests
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero
from scipy.fft import fft, fftfreq

# Abfrage der Nullstellen von LMFDB (hier die ersten 100.000 Nullstellen)
def get_lmfdb_zeros(limit=100000):
    url = f"https://www.lmfdb.org/api/zeros/zeta/?limit={limit}&N=1"
    response = requests.get(url)
    data = response.json()
    return [zero['real'] for zero in data['zeros']]  # Realteil der Nullstellen

# Beispiel: LMFDB-Nullstellen (100.000)
lmfdb_zeros = get_lmfdb_zeros(limit=100000)

# Beispiel-Nullstellen von Odlyzko (erste 10 Nullstellen für die Demonstration)
odlyzko_zeros = [
    14.134725142, 21.022039639, 25.010857580, 30.424876126, 32.935061588,
    37.586178159, 40.918719012, 43.327073281, 48.005150881, 49.773832478
]

# Kombinieren der Nullstellen aus Odlyzko und LMFDB
combined_zeros = odlyzko_zeros + lmfdb_zeros

# Berechnung der Nullstellenabstände
def calculate_distances(zeros):
    return [zeros[i+1] - zeros[i] for i in range(len(zeros) - 1)]

# Berechne die Abstände zwischen den Nullstellen
distances = calculate_distances(combined_zeros)

# Plot der Nullstellenabstände
plt.figure(figsize=(10, 6))
plt.plot(distances, label="Abstände der Nullstellen")
plt.title("Nullstellenabstände der Riemannschen Zetafunktion")
plt.xlabel("Index der Nullstelle")
plt.ylabel("Abstand der Nullstellen")
plt.legend()
plt.grid(True)
plt.show()

# Fibonacci-Skalierung: Bestimmung des Exponenten beta und A
def fibonacci_scaling(distances):
    # Fit auf eine Potenzregel (L(N) = A * N^beta)
    N = np.arange(1, len(distances) + 1)
    log_N = np.log(N)
    log_distances = np.log(distances)
    A, beta = np.polyfit(log_N, log_distances, 1)
    return A, beta

# Bestimme A und Beta für die Fibonacci-Skalierung
A, beta = fibonacci_scaling(distances)

# Ausgabe der berechneten Parameter
print(f"Skalierungsfaktor A: {A}")
print(f"Exponent beta: {beta}")

# Hochrechnen auf größere N (z.B. 10 Millionen Nullstellen)
def scale_to_N(N, A, beta):
    return A * N ** beta

# Berechnung der Kohärenzlängen für größere Nullstellenmengen (z.B. 10 Millionen)
scaled_values = [scale_to_N(N, A, beta) for N in range(1, 11)]

# Plot der hochgerechneten Kohärenzlängen
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), scaled_values, label="Hochgerechnete Kohärenzlängen (bis 10 Millionen)")
plt.title("Hochgerechnete Kohärenzlängen der Nullstellen der Riemannschen Zetafunktion")
plt.xlabel("Anzahl der Nullstellen (N)")
plt.ylabel("Kohärenzlänge (L(N))")
plt.legend()
plt.grid(True)
plt.show()

# Fourier-Analyse zur Überprüfung der Frequenzquantisierung
def fourier_analysis(distances):
    N = len(distances)
    fft_values = fft(distances)
    fft_freqs = fftfreq(N)

    # Berechne den Betrag der Frequenzen
    fft_magnitude = np.abs(fft_values)

    # Plot der Frequenzanalyse
    plt.figure(figsize=(10, 6))
    plt.plot(fft_freqs[:N//2], fft_magnitude[:N//2], label="Fourier-Spektrum")
    plt.title("Fourier-Analyse der Nullstellenabstände")
    plt.xlabel("Frequenz")
    plt.ylabel("Magnitude")
    plt.legend()
    plt.grid(True)
    plt.show()

# Fourier-Analyse der Nullstellenabstände
fourier_analysis(distances)

# Überprüfung der Fibonacci-Skalierung und Frequenzquantisierung
def validate_quantization(beta):
    expected_frequency = np.pi / 8
    print(f"Erwartete Frequenzquantisierung: {expected_frequency}")
    print(f"Berechneter Exponent (beta): {beta}")

    if abs(beta - 1 + (1/np.pi) * (1 + np.sqrt(5))) < 0.01:
        print("Die Fibonacci-Skalierung entspricht den Nullstellenabständen!")
    else:
        print("Es gibt eine Abweichung, die weitere Untersuchung erfordert.")

# Überprüfung der Frequenzquantisierung und Fibonacci-Skalierung
validate_quantization(beta)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter aus der Freese-Formel
A = 0.129376
beta = 0.484964
a = 87.4874
b = -159.7790

# Definition der Freese-Formel
def freese_formula(N, A, beta, a, b):
    return A * N**beta + a * (1 / np.log(N)) - b * np.exp(-np.pi)

# Werte erzeugen
N_values = np.arange(1, 101)
L_N = freese_formula(N_values, A, beta, a, b)

# Zentraler Differenzenoperator (Diskretisierung)
def central_difference_matrix(n):
    D = np.zeros((n, n))
    for i in range(1, n-1):
        D[i, i-1] = -0.5
        D[i, i+1] = 0.5
    return D

# Konstruktion des Operators H_beta
n = len(L_N)
epsilon = 0.0001
D = central_difference_matrix(n)
H_beta = 1j * epsilon * D

# Diagonalisierung
eigenvalues, eigenvectors = np.linalg.eig(H_beta)
eigenvalues_sorted = np.sort(np.real(eigenvalues))

# Plot
plt.figure(figsize=(10, 5))
plt.plot(eigenvalues_sorted, 'o', color='orange')
plt.title("Eigenwerte des Operators $H_\\beta$")
plt.xlabel("Index")
plt.ylabel("Eigenwert (Re)")
plt.grid(True)
plt.show()
```
