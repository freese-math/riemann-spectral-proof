```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib pandas mpmath torch

# 📌 Bibliotheken importieren
import numpy as np
import scipy.special as sp
import pandas as pd
import matplotlib.pyplot as plt
import torch
from mpmath import zetazero, siegeltheta

# 📌 GPU oder CPU wählen
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"🚀 Berechnung auf: {device}")

# 📌 Anzahl der Werte
N = 10000  # Werte für Berechnung

# 📌 1. **Zeta-Nullstellen laden oder berechnen**
try:
    zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:N]
    print(f"✅ {len(zeta_zeros)} Zeta-Nullstellen geladen!")
except:
    print("⚠️ Fehler beim Laden der Datei! Berechne Nullstellen numerisch...")
    zeta_zeros = np.array([float(zetazero(n)) for n in range(1, N+1)])

zeta_zeros = torch.tensor(zeta_zeros, dtype=torch.float32, device=device)

# 📌 2. **Beta-Skala berechnen**
beta_values = torch.log(zeta_zeros) / torch.pi

# 📌 3. **Freese-Reihe generieren**
d = [33, 137]  # Startwerte für Nenner
z = [1, 1]  # Startwerte für Zähler
for n in range(2, N):
    new_d = d[-1] * (3**5 + 9) if n % 2 == 0 else d[-1] * 2 * 3**2
    d.append(new_d)
    z.append(7)

freese_reihe = torch.tensor([z_i / d_i for z_i, d_i in zip(z, d)], dtype=torch.float32, device=device)

# 📌 4. **Siegel-Theta-Funktion berechnen**
t_values = torch.linspace(0.1, 10, N, device=device)
theta_values = torch.tensor([float(siegeltheta(t.item())) for t in t_values], dtype=torch.float32, device=device)

# 📌 5. **Riemannsche Zeta-Funktion evaluieren**
s_values = torch.linspace(0.1, 10, N, device=device)
zeta_values = torch.tensor([float(sp.zeta(s.item())) for s in s_values], dtype=torch.float32, device=device)

# 📌 6. **Euler-Freese-Fehlertest**
euler_freese_values = torch.exp(1j * torch.pi * beta_values) + 1
euler_freese_error = torch.sum(torch.abs(euler_freese_values) ** 2) / N
print(f"✅ Euler-Freese-Fehlerwert: {euler_freese_error:.10f}")

# 📌 7. **Fourier-Analyse-Funktion**
def fourier_analysis(data):
    fft_spectrum = torch.fft.fft(data)
    frequencies = torch.fft.fftfreq(len(data), d=1.0)
    return frequencies, fft_spectrum

# 📌 Fourier-Transformation der Werte
freq_zeta, fft_zeta = fourier_analysis(zeta_zeros)
freq_beta, fft_beta = fourier_analysis(beta_values)
freq_freese, fft_freese = fourier_analysis(freese_reihe)
freq_theta, fft_theta = fourier_analysis(theta_values)
freq_zeta_fn, fft_zeta_fn = fourier_analysis(zeta_values)

# 📌 8. **Dominante Frequenzen extrahieren**
def dominant_freqs(frequencies, fft_spectrum):
    indices = torch.argsort(torch.abs(fft_spectrum))[-5:]
    return frequencies[indices].cpu().numpy(), torch.abs(fft_spectrum[indices]).cpu().numpy()

freq_dominant = {
    "Zeta-Nullstellen": dominant_freqs(freq_zeta, fft_zeta),
    "Beta-Skala": dominant_freqs(freq_beta, fft_beta),
    "Freese-Reihe": dominant_freqs(freq_freese, fft_freese),
    "Siegel-Theta": dominant_freqs(freq_theta, fft_theta),
    "Zeta-Funktion": dominant_freqs(freq_zeta_fn, fft_zeta_fn),
}

# 📌 9. **Ergebnisse ausgeben**
for name, (freqs, amps) in freq_dominant.items():
    print(f"\n🔥 Dominante Frequenzen der {name}: {freqs}")
    print(f"🔥 Korrespondierende Amplituden der {name}: {amps}")

# 📌 10. **Daten in CSV speichern**
df_zeta = pd.DataFrame({"Zeta_Nullstellen": zeta_zeros.cpu().numpy()})
df_beta = pd.DataFrame({"Beta_Skala": beta_values.cpu().numpy()})
df_freese = pd.DataFrame({"Freese_Reihe": freese_reihe.cpu().numpy()})
df_theta = pd.DataFrame({"Siegel_Theta": theta_values.cpu().numpy()})
df_zeta_fn = pd.DataFrame({"Zeta_Funktion": zeta_values.cpu().numpy()})
df_euler_freese = pd.DataFrame({"Euler_Freese": torch.abs(euler_freese_values).cpu().numpy()})

df_zeta.to_csv("/content/zeta_nullstellen.csv", index=False)
df_beta.to_csv("/content/beta_skala.csv", index=False)
df_freese.to_csv("/content/freese_reihe.csv", index=False)
df_theta.to_csv("/content/siegel_theta.csv", index=False)
df_zeta_fn.to_csv("/content/zeta_funktion.csv", index=False)
df_euler_freese.to_csv("/content/euler_freese_test.csv", index=False)

print("✅ Alle Daten erfolgreich als CSV gespeichert!")

# 📌 11. **Plots der Fourier-Spektren**
plt.figure(figsize=(10,5))
plt.plot(freq_zeta.cpu().numpy(), torch.abs(fft_zeta).cpu().numpy(), "b-", label="Zeta-Nullstellen")
plt.plot(freq_beta.cpu().numpy(), torch.abs(fft_beta).cpu().numpy(), "g-", label="Beta-Skala")
plt.plot(freq_freese.cpu().numpy(), torch.abs(fft_freese).cpu().numpy(), "r-", label="Freese-Reihe")
plt.plot(freq_theta.cpu().numpy(), torch.abs(fft_theta).cpu().numpy(), "c-", label="Siegel-Theta")
plt.plot(freq_zeta_fn.cpu().numpy(), torch.abs(fft_zeta_fn).cpu().numpy(), "m-", label="Zeta-Funktion")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektren der mathematischen Reihen")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# ✅ Gegebene Beta-Werte (aus Zeta-Nullstellen abgeleitet)
beta_values = np.random.uniform(0.3, 0.7, 100)  # Beispielhafte Werte
euler_values = np.exp(1j * np.pi * beta_values)  # Euler-Darstellung

# ✅ Cluster-Analyse mit DBSCAN
data = np.column_stack((euler_values.real, euler_values.imag))
dbscan = DBSCAN(eps=0.1, min_samples=2).fit(data)
labels = dbscan.labels_

# ✅ Cluster-Zentren berechnen
unique_labels = set(labels)
cluster_centers = []
for label in unique_labels:
    if label != -1:  # -1 sind Rauschen/Punkte ohne Cluster
        cluster_points = data[labels == label]
        center = cluster_points.mean(axis=0)
        cluster_centers.append(center)

cluster_centers = np.array(cluster_centers)

# ✅ Berechnung der Lücken
angles = np.arctan2(cluster_centers[:, 1], cluster_centers[:, 0])
sorted_indices = np.argsort(angles)
sorted_angles = angles[sorted_indices]

# ✅ Abstände berechnen & größte Lücken markieren
gap_indices = np.argsort(np.diff(sorted_angles))[-33:]  # Größte 33 Lücken

# 🔹 **Visualisierung**
plt.figure(figsize=(6,6))
angles_plot = np.linspace(0, 2*np.pi, 100)
euler_circle = np.exp(1j * angles_plot)

plt.plot(euler_circle.real, euler_circle.imag, 'gray', linestyle="dashed", label="Einheitskreis")
plt.scatter(euler_values.real, euler_values.imag, c=beta_values, cmap="viridis", label=r"$e^{i\beta\pi}$")

# 🔴 Markiere Lücken
for idx in gap_indices:
    gap_x, gap_y = np.cos(sorted_angles[idx]), np.sin(sorted_angles[idx])
    plt.scatter(gap_x, gap_y, color='red', s=80, edgecolor='black', label="Erkannte Lücke" if idx == gap_indices[0] else "")

# ✅ Referenzpunkt Euler-Identität (-1)
plt.scatter([-1], [0], color='red', marker='x', s=100, label="Euler-Identität (-1)")

plt.xlabel("Reelle Achse")
plt.ylabel("Imaginäre Achse")
plt.colorbar(label="Beta-Wert")
plt.title("Erkannte Lücken in $e^{i \ beta\ pi}$")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft
import pywt  # Wavelet-Analyse für tiefere Frequenzanalyse

# 🔹 1. Siegel-Theta-Funktion definieren
def siegel_theta(t, n_max=100):
    """ Berechnet die Siegel-Theta-Funktion als Summenformel. """
    return np.sum(np.exp(-np.pi * n**2 * t) for n in range(1, n_max))

# 🔹 2. Beta(t) aus der Theta-Funktion ableiten
def beta_from_theta(t):
    """ Ableitung von β(t) aus der Siegel-Theta-Funktion. """
    return -np.gradient(siegel_theta(t), t)  # Negative Ableitung als Skalierungsfaktor

# 🔹 3. Wertebereich für t definieren
t_values = np.linspace(0.001, 10, 10000)  # Werte für t (beginnend bei 0.001 für Stabilität)
beta_values_theta = beta_from_theta(t_values)

# 🔹 4. Plot von β(t) aus Siegel-Theta
plt.figure(figsize=(10, 5))
plt.plot(t_values, beta_values_theta, 'b-', label=r'$\beta(t)$ aus Siegel-Theta')
plt.xlabel("t")
plt.ylabel(r"$\beta(t)$")
plt.title(r"Direkte Ableitung von $\beta(t)$ aus Siegel-Theta")
plt.legend()
plt.grid()
plt.show()

# 🔹 5. Fourier-Analyse von β(t)
beta_fft = fft.fft(beta_values_theta)
frequencies = fft.fftfreq(len(t_values), d=(t_values[1] - t_values[0]))

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(beta_fft[:len(beta_fft)//2]), 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von β(t) aus Siegel-Theta")
plt.grid()
plt.show()

# 🔹 6. Wavelet-Analyse
scales = np.arange(1, 500)
coefficients, frequencies = pywt.cwt(beta_values_theta, scales, 'cmor')

plt.figure(figsize=(10, 5))
plt.imshow(np.abs(coefficients), aspect='auto', extent=[t_values.min(), t_values.max(), scales.max(), scales.min()], cmap='viridis')
plt.colorbar(label="Magnitude")
plt.xlabel("t")
plt.ylabel("Skalierung")
plt.title("Wavelet-Analyse (CWT) von β(t) aus Siegel-Theta")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import pywt  # Für Wavelet-Analyse

# 🔹 1. Laden der Zeta-Nullstellen
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
zeta_zeros = np.loadtxt(url)

# 🔹 2. Berechnung von β(t) aus den Zeta-Nullstellen
def beta_from_zeros(zeros):
    """ Berechnet β(t) basierend auf den Abständen der Zeta-Nullstellen. """
    t_values = np.diff(zeros)  # Differenzen zwischen aufeinanderfolgenden Nullstellen
    beta_values = 1 / t_values  # Inverse der Abstände als Näherung für β(t)
    return beta_values

beta_values = beta_from_zeros(zeta_zeros)

# 🔹 3. Fourier-Analyse von β(t)
beta_fft = fft.fft(beta_values)
frequencies = fft.fftfreq(len(beta_values), d=1)

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(beta_fft[:len(beta_fft)//2]), 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von β(t) aus Zeta-Nullstellen")
plt.grid()
plt.show()

# 🔹 4. Wavelet-Analyse
scales = np.arange(1, 500)
coefficients, frequencies = pywt.cwt(beta_values, scales, 'cmor')

plt.figure(figsize=(10, 5))
plt.imshow(np.abs(coefficients), aspect='auto', extent=[0, len(beta_values), scales.max(), scales.min()], cmap='viridis')
plt.colorbar(label="Magnitude")
plt.xlabel("Index der Nullstelle")
plt.ylabel("Skalierung")
plt.title("Wavelet-Analyse (CWT) von β(t) aus Zeta-Nullstellen")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft

# ✅ Berechnung der Siegel-Theta-Funktion
def siegel_theta(t, n_max=100):
    return np.sum(np.exp(-np.pi * n**2 * t) for n in range(1, n_max))

# ✅ Berechnung von \(\beta(t)\) als Korrekturterm
def beta_function(t, A, C, gamma):
    return A + C * t**(-gamma)

# ✅ Werte für t
t_values = np.linspace(0.01, 10, 10000)
theta_values = np.array([siegel_theta(t) for t in t_values])

# ✅ Vergleich mit Skalierungsrelation
beta_values = theta_values / t_values**(1/4)  # Ansatz für β(t) als Modulation

# ✅ Fit von β(t) an eine allgemeine Skalierungsform
popt, _ = opt.curve_fit(beta_function, t_values, beta_values, p0=[0.5, 0.4, 0.2])

A_opt, C_opt, gamma_opt = popt

# ✅ Plots der Ergebnisse
plt.figure(figsize=(10, 5))
plt.plot(t_values, beta_values, 'r.', markersize=1, label="Numerische $\beta(t)$-Werte")
plt.plot(t_values, beta_function(t_values, *popt), 'b-', label=f"Fit: A={A_opt:.5f}, C={C_opt:.5f}, γ={gamma_opt:.5f}")
plt.xlabel("$t$")
plt.ylabel("$\\beta(t)$")
plt.title("Direkte Ableitung von $\beta(t)$ aus Siegel-Theta")
plt.legend()
plt.grid()
plt.show()

# ✅ Fourier-Analyse von β(t)
frequencies = fft.fftfreq(len(t_values), d=(t_values[1] - t_values[0]))
amplitudes = np.abs(fft.fft(beta_values))

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], amplitudes[:len(amplitudes)//2], 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von $\beta(t)$ aus Siegel-Theta")
plt.grid()
plt.show()

# ✅ Optimierte Werte ausgeben
print(f"Optimierte Werte: A={A_opt:.5f}, C={C_opt:.5f}, γ={gamma_opt:.5f}")
```
