```python
# üìå Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib pandas mpmath torch

# üìå Bibliotheken importieren
import numpy as np
import scipy.special as sp
import pandas as pd
import matplotlib.pyplot as plt
import torch
from mpmath import zetazero, siegeltheta

# üìå GPU oder CPU w√§hlen
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"üöÄ Berechnung auf: {device}")

# üìå Anzahl der Werte
N = 10000  # Werte f√ºr Berechnung

# üìå 1. **Zeta-Nullstellen laden oder berechnen**
try:
    zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:N]
    print(f"‚úÖ {len(zeta_zeros)} Zeta-Nullstellen geladen!")
except:
    print("‚ö†Ô∏è Fehler beim Laden der Datei! Berechne Nullstellen numerisch...")
    zeta_zeros = np.array([float(zetazero(n)) for n in range(1, N+1)])

zeta_zeros = torch.tensor(zeta_zeros, dtype=torch.float32, device=device)

# üìå 2. **Beta-Skala berechnen**
beta_values = torch.log(zeta_zeros) / torch.pi

# üìå 3. **Freese-Reihe generieren**
d = [33, 137]  # Startwerte f√ºr Nenner
z = [1, 1]  # Startwerte f√ºr Z√§hler
for n in range(2, N):
    new_d = d[-1] * (3**5 + 9) if n % 2 == 0 else d[-1] * 2 * 3**2
    d.append(new_d)
    z.append(7)

freese_reihe = torch.tensor([z_i / d_i for z_i, d_i in zip(z, d)], dtype=torch.float32, device=device)

# üìå 4. **Siegel-Theta-Funktion berechnen**
t_values = torch.linspace(0.1, 10, N, device=device)
theta_values = torch.tensor([float(siegeltheta(t.item())) for t in t_values], dtype=torch.float32, device=device)

# üìå 5. **Riemannsche Zeta-Funktion evaluieren**
s_values = torch.linspace(0.1, 10, N, device=device)
zeta_values = torch.tensor([float(sp.zeta(s.item())) for s in s_values], dtype=torch.float32, device=device)

# üìå 6. **Euler-Freese-Fehlertest**
euler_freese_values = torch.exp(1j * torch.pi * beta_values) + 1
euler_freese_error = torch.sum(torch.abs(euler_freese_values) ** 2) / N
print(f"‚úÖ Euler-Freese-Fehlerwert: {euler_freese_error:.10f}")

# üìå 7. **Fourier-Analyse-Funktion**
def fourier_analysis(data):
    fft_spectrum = torch.fft.fft(data)
    frequencies = torch.fft.fftfreq(len(data), d=1.0)
    return frequencies, fft_spectrum

# üìå Fourier-Transformation der Werte
freq_zeta, fft_zeta = fourier_analysis(zeta_zeros)
freq_beta, fft_beta = fourier_analysis(beta_values)
freq_freese, fft_freese = fourier_analysis(freese_reihe)
freq_theta, fft_theta = fourier_analysis(theta_values)
freq_zeta_fn, fft_zeta_fn = fourier_analysis(zeta_values)

# üìå 8. **Dominante Frequenzen extrahieren**
def dominant_freqs(frequencies, fft_spectrum):
    indices = torch.argsort(torch.abs(fft_spectrum))[-5:]
    return frequencies[indices].cpu().numpy(), torch.abs(fft_spectrum[indices]).cpu().numpy()

freq_dominant = {
    "Zeta-Nullstellen": dominant_freqs(freq_zeta, fft_zeta),
    "Beta-Skala": dominant_freqs(freq_beta, fft_beta),
    "Freese-Reihe": dominant_freqs(freq_freese, fft_freese),
    "Siegel-Theta": dominant_freqs(freq_theta, fft_theta),
    "Zeta-Funktion": dominant_freqs(freq_zeta_fn, fft_zeta_fn),
}

# üìå 9. **Ergebnisse ausgeben**
for name, (freqs, amps) in freq_dominant.items():
    print(f"\nüî• Dominante Frequenzen der {name}: {freqs}")
    print(f"üî• Korrespondierende Amplituden der {name}: {amps}")

# üìå 10. **Daten in CSV speichern**
df_zeta = pd.DataFrame({"Zeta_Nullstellen": zeta_zeros.cpu().numpy()})
df_beta = pd.DataFrame({"Beta_Skala": beta_values.cpu().numpy()})
df_freese = pd.DataFrame({"Freese_Reihe": freese_reihe.cpu().numpy()})
df_theta = pd.DataFrame({"Siegel_Theta": theta_values.cpu().numpy()})
df_zeta_fn = pd.DataFrame({"Zeta_Funktion": zeta_values.cpu().numpy()})
df_euler_freese = pd.DataFrame({"Euler_Freese": torch.abs(euler_freese_values).cpu().numpy()})

df_zeta.to_csv("/content/zeta_nullstellen.csv", index=False)
df_beta.to_csv("/content/beta_skala.csv", index=False)
df_freese.to_csv("/content/freese_reihe.csv", index=False)
df_theta.to_csv("/content/siegel_theta.csv", index=False)
df_zeta_fn.to_csv("/content/zeta_funktion.csv", index=False)
df_euler_freese.to_csv("/content/euler_freese_test.csv", index=False)

print("‚úÖ Alle Daten erfolgreich als CSV gespeichert!")

# üìå 11. **Plots der Fourier-Spektren**
plt.figure(figsize=(10,5))
plt.plot(freq_zeta.cpu().numpy(), torch.abs(fft_zeta).cpu().numpy(), "b-", label="Zeta-Nullstellen")
plt.plot(freq_beta.cpu().numpy(), torch.abs(fft_beta).cpu().numpy(), "g-", label="Beta-Skala")
plt.plot(freq_freese.cpu().numpy(), torch.abs(fft_freese).cpu().numpy(), "r-", label="Freese-Reihe")
plt.plot(freq_theta.cpu().numpy(), torch.abs(fft_theta).cpu().numpy(), "c-", label="Siegel-Theta")
plt.plot(freq_zeta_fn.cpu().numpy(), torch.abs(fft_zeta_fn).cpu().numpy(), "m-", label="Zeta-Funktion")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensit√§t")
plt.title("Fourier-Spektren der mathematischen Reihen")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from sklearn.cluster import DBSCAN

# ‚úÖ Zeta-Nullstellen von der Odlyzko-Seite laden
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(url)
data_text = response.text.splitlines()

# ‚úÖ Daten in Float umwandeln und fehlerhafte Werte entfernen
zeta_zeros = np.array([float(x) for x in data_text if x.strip()])
zeta_zeros = zeta_zeros[~np.isnan(zeta_zeros) & ~np.isinf(zeta_zeros)]  # Entfernt NaN & inf

# ‚úÖ N kleiner w√§hlen f√ºr RAM-Optimierung
N = np.arange(1, 1000)  # üîπ Reduziert von 10.000 auf 1.000 f√ºr Speicher-Effizienz

# ‚úÖ Speicherfreundliche Berechnung der Beta-Skala
def compute_beta(N, zeta_zeros, batch_size=50):
    beta = np.zeros_like(N, dtype=np.float64)
    log_N = np.log(N)
    log_N[log_N == 0] = 1e-10  # üîπ Vermeidet Division durch 0

    num_batches = len(zeta_zeros) // batch_size
    for i in range(num_batches + 1):
        batch = zeta_zeros[i * batch_size:(i + 1) * batch_size]
        if len(batch) == 0:
            continue
        beta += np.sum(np.cos(batch * log_N[:, None]), axis=1)

    return 0.5 - (1 / log_N) * beta  # üîπ Skaliert nach Freese-Formel

# ‚úÖ Beta-Werte berechnen
beta_values = compute_beta(N, zeta_zeros)

# ‚úÖ Euler-Kreisdarstellung
euler_values = np.exp(1j * np.pi * beta_values)

# ‚úÖ Cluster-Analyse mit DBSCAN
data = np.column_stack((euler_values.real, euler_values.imag))
dbscan = DBSCAN(eps=0.05, min_samples=3).fit(data)
labels = dbscan.labels_

# ‚úÖ Cluster-Zentren berechnen
unique_labels = set(labels)
cluster_centers = []
for label in unique_labels:
    if label != -1:
        cluster_points = data[np.where(labels == label)[0]]
        cluster_centers.append(cluster_points.mean(axis=0))

cluster_centers = np.array(cluster_centers)

# ‚úÖ Berechnung der L√ºcken
if len(cluster_centers) > 0:
    angles = np.arctan2(cluster_centers[:, 1], cluster_centers[:, 0])
    sorted_indices = np.argsort(angles)
    sorted_angles = angles[sorted_indices]

    # ‚úÖ Gr√∂√üte L√ºcken markieren (Falls weniger als 33 existieren, dann alle)
    num_gaps = min(33, len(sorted_angles) - 1)
    gap_indices = np.argsort(np.diff(sorted_angles))[-num_gaps:]

# üîπ **Visualisierung**
plt.figure(figsize=(6,6))
angles_plot = np.linspace(0, 2*np.pi, 100)
euler_circle = np.exp(1j * angles_plot)

plt.plot(euler_circle.real, euler_circle.imag, 'gray', linestyle="dashed", label="Einheitskreis")
plt.scatter(euler_values.real, euler_values.imag, c=beta_values, cmap="viridis", label=r"$e^{i\beta\pi}$")

# üî¥ Markiere L√ºcken
if len(cluster_centers) > 0:
    for idx in gap_indices:
        gap_x, gap_y = np.cos(sorted_angles[idx]), np.sin(sorted_angles[idx])
        plt.scatter(gap_x, gap_y, color='red', s=80, edgecolor='black', label="Erkannte L√ºcke" if idx == gap_indices[0] else "")

# ‚úÖ Referenzpunkt Euler-Identit√§t (-1)
plt.scatter([-1], [0], color='red', marker='x', s=100, label="Euler-Identit√§t (-1)")

plt.xlabel("Reelle Achse")
plt.ylabel("Imagin√§re Achse")
plt.colorbar(label="Beta-Wert")
plt.title("Erkannte L√ºcken in $e^{i \ beta\ pi}$ aus Zeta-Nullstellen")
plt.legend()
plt.grid()
plt.show()
```

```python
# üîπ 1. Wichtige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
from scipy.fftpack import fft, fftfreq
from scipy.special import zeta
import requests

# üîπ 2. Siegel-Theta-Funktion definieren
def siegel_theta(t, n_max=50):
    return np.sum(np.exp(-np.pi * n**2 * t) for n in range(1, n_max))

# üîπ 3. Beta-Formel f√ºr Zeta-Nullstellen
def beta_function(t, A, C, gamma):
    return A + C * t**(-gamma)

# üîπ 4. Zeta-Nullstellen von Odlyzko laden
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(url)
zeta_zeros = np.array([float(line.strip()) for line in response.text.split("\n") if line.strip()])

# üîπ 5. Berechnung von Œ≤(t) aus Zeta-Nullstellen
t_values = zeta_zeros[:10000]  # Erste 10.000 Nullstellen
beta_values_zeta = -np.pi * t_values / np.log(t_values)  # Theorieformel

# üîπ 6. Curve Fit zur Bestimmung von A, C, Œ≥ f√ºr Œ≤ aus Zeta
popt_zeta, _ = opt.curve_fit(beta_function, t_values, beta_values_zeta, p0=[0.5, 0.4, 0.2])
A_z, C_z, gamma_z = popt_zeta

# üîπ 7. Berechnung von Œ≤(t) aus Siegel-Theta
theta_values = np.array([siegel_theta(t) for t in t_values])
beta_values_theta = np.gradient(theta_values, t_values)  # Ableitung von Œò(t)

# üîπ 8. Curve Fit f√ºr Œ≤(t) aus Siegel-Theta
popt_theta, _ = opt.curve_fit(beta_function, t_values, beta_values_theta, p0=[0.5, 0.4, 0.2])
A_t, C_t, gamma_t = popt_theta

# üîπ 9. Fourier-Analyse von Œ≤(t) aus Zeta & Theta
beta_fft_zeta = fft(beta_values_zeta)
beta_fft_theta = fft(beta_values_theta)
frequencies = fftfreq(len(t_values), d=(t_values[1] - t_values[0]))

# üîπ 10. Visualisierung der Ergebnisse
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# üîπ Plot 1: Vergleich Œ≤(t) aus Zeta und Siegel-Theta
axes[0].plot(t_values, beta_values_zeta, 'r.', markersize=1, label="Œ≤ aus Zeta-Nullstellen")
axes[0].plot(t_values, beta_values_theta, 'b.', markersize=1, label="Œ≤ aus Siegel-Theta")
axes[0].set_title("Vergleich: Œ≤(t) aus Zeta vs. Œ≤(t) aus Siegel-Theta")
axes[0].legend()

# üîπ Plot 2: Fourier-Analyse von Œ≤(t)
axes[1].plot(frequencies[:len(frequencies)//2], np.abs(beta_fft_zeta[:len(frequencies)//2]), 'r', label="Œ≤ aus Zeta")
axes[1].plot(frequencies[:len(frequencies)//2], np.abs(beta_fft_theta[:len(frequencies)//2]), 'b', label="Œ≤ aus Theta")
axes[1].set_title("Fourier-Analyse von Œ≤(t)")
axes[1].legend()

# üîπ Plot 3: Fourier-Analyse von e^(iŒ≤œÄ) + 1
e_beta_zeta = np.exp(1j * beta_values_zeta * np.pi) + 1
e_beta_theta = np.exp(1j * beta_values_theta * np.pi) + 1
e_beta_fft_zeta = fft(np.real(e_beta_zeta))
e_beta_fft_theta = fft(np.real(e_beta_theta))
axes[2].plot(frequencies[:len(frequencies)//2], np.abs(e_beta_fft_zeta[:len(frequencies)//2]), 'r', label="e^(iŒ≤œÄ) + 1 (Zeta)")
axes[2].plot(frequencies[:len(frequencies)//2], np.abs(e_beta_fft_theta[:len(frequencies)//2]), 'b', label="e^(iŒ≤œÄ) + 1 (Theta)")
axes[2].set_title("Fourier-Analyse von e^(iŒ≤œÄ) + 1")
axes[2].legend()

plt.tight_layout()
plt.show()

# üöÄ Ergebnis:
print(f"Optimierte Werte f√ºr Œ≤ aus Zeta: A={A_z:.5f}, C={C_z:.5f}, Œ≥={gamma_z:.5f}")
print(f"Optimierte Werte f√ºr Œ≤ aus Theta: A={A_t:.5f}, C={C_t:.5f}, Œ≥={gamma_t:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft
import pywt  # Wavelet-Analyse f√ºr tiefere Frequenzanalyse

# üîπ 1. Siegel-Theta-Funktion definieren
def siegel_theta(t, n_max=100):
    """ Berechnet die Siegel-Theta-Funktion als Summenformel. """
    return np.sum(np.exp(-np.pi * n**2 * t) for n in range(1, n_max))

# üîπ 2. Beta(t) aus der Theta-Funktion ableiten
def beta_from_theta(t):
    """ Ableitung von Œ≤(t) aus der Siegel-Theta-Funktion. """
    return -np.gradient(siegel_theta(t), t)  # Negative Ableitung als Skalierungsfaktor

# üîπ 3. Wertebereich f√ºr t definieren
t_values = np.linspace(0.001, 10, 10000)  # Werte f√ºr t (beginnend bei 0.001 f√ºr Stabilit√§t)
beta_values_theta = beta_from_theta(t_values)

# üîπ 4. Plot von Œ≤(t) aus Siegel-Theta
plt.figure(figsize=(10, 5))
plt.plot(t_values, beta_values_theta, 'b-', label=r'$\beta(t)$ aus Siegel-Theta')
plt.xlabel("t")
plt.ylabel(r"$\beta(t)$")
plt.title(r"Direkte Ableitung von $\beta(t)$ aus Siegel-Theta")
plt.legend()
plt.grid()
plt.show()

# üîπ 5. Fourier-Analyse von Œ≤(t)
beta_fft = fft.fft(beta_values_theta)
frequencies = fft.fftfreq(len(t_values), d=(t_values[1] - t_values[0]))

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(beta_fft[:len(beta_fft)//2]), 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von Œ≤(t) aus Siegel-Theta")
plt.grid()
plt.show()

# üîπ 6. Wavelet-Analyse
scales = np.arange(1, 500)
coefficients, frequencies = pywt.cwt(beta_values_theta, scales, 'cmor')

plt.figure(figsize=(10, 5))
plt.imshow(np.abs(coefficients), aspect='auto', extent=[t_values.min(), t_values.max(), scales.max(), scales.min()], cmap='viridis')
plt.colorbar(label="Magnitude")
plt.xlabel("t")
plt.ylabel("Skalierung")
plt.title("Wavelet-Analyse (CWT) von Œ≤(t) aus Siegel-Theta")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import pywt

# üîπ 1. Laden der Zeta-Nullstellen (nur die ersten 100.000)
url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
zeta_zeros = np.loadtxt(url, max_rows=100000)

# üîπ 2. Berechnung von Œ≤(t) aus den Nullstellen (Differenzen zur Normalisierung)
def beta_from_zeros(zeros):
    t_values = np.diff(zeros)
    beta_values = 1 / t_values
    return beta_values - np.mean(beta_values)  # Zentrierung um den Mittelwert

beta_values = beta_from_zeros(zeta_zeros)

# üîπ 3. Fourier-Analyse (FFT auf normalisierte Daten)
beta_fft = fft.fft(beta_values)
frequencies = fft.fftfreq(len(beta_values), d=1)

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(beta_fft[:len(beta_fft)//2]), 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Optimierte Fourier-Analyse von Œ≤(t) aus Zeta-Nullstellen")
plt.grid()
plt.show()

# üîπ 4. Wavelet-Analyse (mit optimierten Skalen)
scales = np.arange(1, 300)
coefficients, frequencies = pywt.cwt(beta_values, scales, 'cmor')

plt.figure(figsize=(10, 5))
plt.imshow(np.abs(coefficients), aspect='auto', extent=[0, len(beta_values), scales.max(), scales.min()], cmap='viridis')
plt.colorbar(label="Magnitude")
plt.xlabel("Index der Nullstelle")
plt.ylabel("Skalierung")
plt.title("Optimierte Wavelet-Analyse (CWT) von Œ≤(t) aus Zeta-Nullstellen")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft

# ‚úÖ Berechnung der Siegel-Theta-Funktion
def siegel_theta(t, n_max=100):
    return np.sum(np.exp(-np.pi * n**2 * t) for n in range(1, n_max))

# ‚úÖ Berechnung von \(\beta(t)\) als Korrekturterm
def beta_function(t, A, C, gamma):
    return A + C * t**(-gamma)

# ‚úÖ Werte f√ºr t
t_values = np.linspace(0.01, 10, 10000)
theta_values = np.array([siegel_theta(t) for t in t_values])

# ‚úÖ Vergleich mit Skalierungsrelation
beta_values = theta_values / t_values**(1/4)  # Ansatz f√ºr Œ≤(t) als Modulation

# ‚úÖ Fit von Œ≤(t) an eine allgemeine Skalierungsform
popt, _ = opt.curve_fit(beta_function, t_values, beta_values, p0=[0.5, 0.4, 0.2])

A_opt, C_opt, gamma_opt = popt

# ‚úÖ Plots der Ergebnisse
plt.figure(figsize=(10, 5))
plt.plot(t_values, beta_values, 'r.', markersize=1, label="Numerische $\beta(t)$-Werte")
plt.plot(t_values, beta_function(t_values, *popt), 'b-', label=f"Fit: A={A_opt:.5f}, C={C_opt:.5f}, Œ≥={gamma_opt:.5f}")
plt.xlabel("$t$")
plt.ylabel("$\\beta(t)$")
plt.title("Direkte Ableitung von $\beta(t)$ aus Siegel-Theta")
plt.legend()
plt.grid()
plt.show()

# ‚úÖ Fourier-Analyse von Œ≤(t)
frequencies = fft.fftfreq(len(t_values), d=(t_values[1] - t_values[0]))
amplitudes = np.abs(fft.fft(beta_values))

plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], amplitudes[:len(amplitudes)//2], 'b')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse von $\beta(t)$ aus Siegel-Theta")
plt.grid()
plt.show()

# ‚úÖ Optimierte Werte ausgeben
print(f"Optimierte Werte: A={A_opt:.5f}, C={C_opt:.5f}, Œ≥={gamma_opt:.5f}")
```
