```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange, log as sympy_log
from math import log
from tqdm import tqdm

# -- Konfiguration --
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
N = 1000         # Bereich der n-Werte
M = 500          # Anzahl Beta-Werte im Scan
beta_grid = torch.linspace(0.1, 2.0, M, device=device)

# -- Lambda(n): Mangoldt-Approx ber log(p), sonst 0 --
Lambda_n = torch.zeros(N, device=device)
primes = list(primerange(1, N + 1))
for p in primes:
    k = p
    while k <= N:
        Lambda_n[k - 1] = log(p)
        k *= p

# -- Heatmap-Array --
entropy_map = torch.zeros((M, N), device=device)

# -- Entropie berechnen --
for j, beta in enumerate(tqdm(beta_grid)):
    weights = Lambda_n / torch.pow(torch.arange(1, N + 1, device=device), beta)
    for i in range(1, N + 1):
        local = weights[:i]
        prob = local / torch.sum(local)
        prob = prob[prob > 0]
        entropy = -torch.sum(prob * torch.log(prob))
        entropy_map[j, i - 1] = entropy

# -- (n) = argmin Entropie --
min_entropy_vals, min_entropy_idx = torch.min(entropy_map, dim=0)
beta_n = beta_grid[min_entropy_idx].cpu().numpy()

# -- Plot Heatmap --
plt.figure(figsize=(14, 5))
plt.imshow(entropy_map.cpu().numpy(), aspect='auto', origin='lower',
           extent=[1, N, beta_grid[0].item(), beta_grid[-1].item()], cmap='viridis')
plt.colorbar(label='Entropie')
plt.xlabel("n")
plt.ylabel(r"$\beta$")
plt.title(r"Heatmap der Entropie $H(\beta,n)$ zur Optimierung von $\beta(n)$")
plt.tight_layout()
plt.show()

# -- Plot (n) Verlauf --
plt.figure(figsize=(14, 4))
plt.plot(beta_n, color='darkred')
plt.title(r"Lokal optimiertes $\beta(n)$ aus Entropie-Minimierung")
plt.xlabel("n")
plt.ylabel(r"$\beta(n)$")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ---------------------------------------
# Vorbereitungen
# ---------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import torch
from torch.fft import fft
from sympy import primerange
from sympy.functions.combinatorial.numbers import mobius

# Parameter
N = 1000
beta_grid = np.linspace(0.1, 2.0, 100)
n_grid = np.arange(1, N + 1)

# Lambda-Funktion berechnen (Mangoldt)
def mangoldt(n):
    from sympy import factorint
    f = factorint(n)
    return np.log(list(f.keys())[0]) if len(f) == 1 and list(f.values())[0] == 1 else 0

Lambda = np.array([mangoldt(n) for n in n_grid])

# ---------------------------------------
# Entropie-Matrix berechnen
# ---------------------------------------
H = np.zeros((len(beta_grid), len(n_grid)))

for i, beta in enumerate(beta_grid):
    weights = Lambda / (n_grid ** beta)
    weights = np.abs(weights)
    normed = weights / np.sum(weights) if np.sum(weights) > 0 else np.zeros_like(weights)
    H[i, :] = -np.sum(normed * np.log(normed + 1e-12))  # Entropieformel

# ---------------------------------------
# 3D-Surface Plot
# ---------------------------------------
B, Nn = np.meshgrid(n_grid, beta_grid)
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(B, Nn, H, cmap='viridis')
ax.set_xlabel('n')
ax.set_ylabel(r'$\beta$')
ax.set_zlabel('Entropie')
ax.set_title(r'3D-Oberflche der Entropie $H(\beta, n)$')
plt.colorbar(surf, shrink=0.5)
plt.show()

# ---------------------------------------
# Lokales (n) aus minimaler Entropie extrahieren
# ---------------------------------------
beta_opt = beta_grid[np.argmin(H, axis=0)]

# ---------------------------------------
# Fourier-Analyse der gewichteten Terme
# ---------------------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
n_tensor = torch.arange(1, N + 1, device=device, dtype=torch.float32)
Lambda_tensor = torch.tensor(Lambda, device=device, dtype=torch.float32)
beta_tensor = torch.tensor(beta_opt, device=device, dtype=torch.float32)

weighted_terms = Lambda_tensor / torch.pow(n_tensor, beta_tensor)
fft_result = fft(weighted_terms)
frequencies = np.fft.fftfreq(N, d=1)
amplitudes = torch.abs(fft_result).cpu().numpy()

# ---------------------------------------
# Plot Fourier-Spektrum
# ---------------------------------------
plt.figure(figsize=(12, 4))
plt.plot(frequencies[:N//2], amplitudes[:N//2], color='purple')
plt.title(r'Fourier-Spektrum der gewichteten Mangoldt-Terme $\Lambda(n)/n^{\beta(n)}$')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange
import pandas as pd

# Pfade zu den Skalen
file_beta_kombi = '/content/drive/MyDrive/beta_skala_kombi.csv'

# --- Parameter ---
N = 10_000  # Anzahl Terme

# --- Mangoldt-Funktion ---
def mangoldt(n):
    from sympy import factorint
    if n < 2:
        return 0
    f = factorint(n)
    if len(f) == 1 and list(f.values())[0] == 1:
        return np.log(list(f.keys())[0])
    return 0

# --- Beta(n) laden ---
beta_vals = pd.read_csv(file_beta_kombi, header=None).squeeze().values
beta_vals = beta_vals[:N]

# Problematische Beta-Werte korrigieren
beta_vals = np.clip(beta_vals, 0.01, 2.0)

# --- Lambda(n) berechnen ---
Lambda_n = np.array([mangoldt(n) for n in range(1, N + 1)])

# --- Logarithmisch stabilisiert: Lambda(n) / n^(n) = Lambda(n) * exp(-(n) * log(n)) ---
n_values = np.arange(1, N + 1)
log_n = np.log(n_values)
weighted_terms = Lambda_n * np.exp(-beta_vals * log_n)

# --- Kumulative Summe ---
cumulative_sum = np.cumsum(weighted_terms)

# --- Plot ---
plt.figure(figsize=(12, 5))
plt.plot(n_values, cumulative_sum, color='orange', alpha=0.8, label=r'$\sum_{k=1}^n \frac{\Lambda(k)}{k^{\beta(k)}}$')
plt.axhline(0, linestyle='--', color='gray', linewidth=0.8)
plt.title('Kumulative gewichtete Mangoldt-Summe mit (n)')
plt.xlabel('n')
plt.ylabel(r'$\sum_{k=1}^n \frac{\Lambda(k)}{k^{\beta(k)}}$')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.stats import pearsonr

# Gert whlen
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Verwende Gert: {device}")

# === Datei-Pfade ===
prime_file = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
beta_op_file = '/content/drive/MyDrive/beta_scale_operator.csv'
beta_kom_file = '/content/drive/MyDrive/beta_skala_kombi.csv'
beta_prz_file = '/content/drive/MyDrive/beta_skala_przision.csv'
zeta_file = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt'

# === Primzahlen laden ===
prime_data = pd.read_csv(prime_file, header=None)
prime_data = prime_data[pd.to_numeric(prime_data[0], errors='coerce').notna()]
primes = prime_data[0].astype(int).values
print(f"Primzahlen geladen: {len(primes)}")

N = len(primes)
n = torch.arange(1, N + 1, dtype=torch.float32, device=device)

# === Beta-Skalen laden ===
def load_beta(file, col=0):
    df = pd.read_csv(file, header=None)
    df = df[pd.to_numeric(df[col], errors='coerce').notna()]
    return torch.tensor(df[col].astype(float).values[:N], dtype=torch.float32, device=device)

beta_op = load_beta(beta_op_file, col=1)
beta_kom = load_beta(beta_kom_file)
beta_prz = load_beta(beta_prz_file)

# === Lambda(n) bauen ===
Lambda_np = np.zeros(N, dtype=np.float32)
for p in primes:
    if p <= N:
        Lambda_np[p - 1] = np.log(p)
Lambda_n = torch.tensor(Lambda_np, dtype=torch.float32, device=device)

# === Terme berechnen ===
def weighted_terms(beta): return Lambda_n / torch.pow(n, beta)

terms_op = weighted_terms(beta_op)
terms_kom = weighted_terms(beta_kom)
terms_prz = weighted_terms(beta_prz)

# === Summenvergleich ===
print(f"Summe _op: {terms_op.sum().item():.4f}")
print(f"Summe _kom: {terms_kom.sum().item():.4f}")
print(f"Summe _prz: {terms_prz.sum().item():.4f}")

# === Plot: gewichtete Terme ===
plt.figure(figsize=(14,5))
plt.plot(n.cpu(), terms_op.cpu(), label="Operator", alpha=0.6)
plt.plot(n.cpu(), terms_kom.cpu(), label="Kombi", alpha=0.6)
plt.plot(n.cpu(), terms_prz.cpu(), label="Przision", alpha=0.6)
plt.legend(); plt.title("Gewichtete Euler-Freese-Terme"); plt.xlabel("n"); plt.ylabel("Wert")
plt.grid(True); plt.tight_layout(); plt.show()

# === Fourier-Analyse ===
def fourier_plot(terms, label):
    Y = fft(terms.cpu().numpy())
    freqs = fftfreq(len(Y))
    plt.plot(freqs[:len(freqs)//2], np.abs(Y[:len(Y)//2]), label=label)

plt.figure(figsize=(14,4))
fourier_plot(terms_op, "Operator")
fourier_plot(terms_kom, "Kombi")
fourier_plot(terms_prz, "Przision")
plt.title("Fourier-Spektren der gewichteten Terme")
plt.xlabel("Frequenz"); plt.ylabel("Amplitude"); plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

# === Zeta-Korrelation ===
zeta_zeros = pd.read_csv(zeta_file, header=None)[0].dropna().astype(float).values[:N]
for name, terms in [("Operator", terms_op), ("Kombi", terms_kom), ("Przision", terms_prz)]:
    r = pearsonr(terms[:N].cpu().numpy(), zeta_zeros)[0]
    print(f"Korrelation mit Zeta-Nullstellen ({name}): {r:.5f}")

# === Heatmap von (n) ===
beta_stack = torch.stack([beta_op, beta_kom, beta_prz])
labels = ['Operator', 'Kombi', 'Przision']

plt.figure(figsize=(12,6))
plt.imshow(beta_stack.cpu().numpy(), aspect='auto', cmap='plasma')
plt.colorbar(label="(n)")
plt.yticks(ticks=[0,1,2], labels=labels)
plt.title("Heatmap der Beta-Skalen")
plt.xlabel("n")
plt.tight_layout(); plt.show()

# === Lokale gleitende Fenster-Mittelwerte ===
def rolling_avg(tensor, window=10000):
    return np.convolve(tensor.cpu().numpy(), np.ones(window)/window, mode='valid')

plt.figure(figsize=(14,5))
plt.plot(rolling_avg(beta_op), label='Operator', color='navy')
plt.plot(rolling_avg(beta_kom), label='Kombi', color='orange')
plt.plot(rolling_avg(beta_prz), label='Przision', color='green')
plt.title("Gleitender Mittelwert von (n)  Fenstergre 10.000")
plt.xlabel("n"); plt.ylabel("(n) Mittel")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Verwende Gert: {device}")

# === 1. Pfade ===
prime_file = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
zeta_file = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt'
beta_op_file = '/content/drive/MyDrive/beta_scale_operator.csv'
beta_kom_file = '/content/drive/MyDrive/beta_skala_kombi.csv'
beta_prz_file = '/content/drive/MyDrive/beta_skala_przision.csv'

# === 2. Primzahlen laden ===
prime_data = pd.read_csv(prime_file, header=None)
prime_data = prime_data[pd.to_numeric(prime_data[0], errors='coerce').notna()]
primes = prime_data[0].astype(int).values

# === 3. Beta-Skalen laden ===
beta_op_df = pd.read_csv(beta_op_file, header=0)
beta_kom_df = pd.read_csv(beta_kom_file, header=0)
beta_prz_df = pd.read_csv(beta_prz_file, header=None, skiprows=1)

# Spalten bereinigen und konvertieren
beta_op_vals = pd.to_numeric(beta_op_df.iloc[:, 1], errors='coerce').dropna().values
beta_kom_vals = pd.to_numeric(beta_kom_df.iloc[:, 0], errors='coerce').dropna().values
beta_prz_vals = pd.to_numeric(beta_prz_df[0], errors='coerce').dropna().values

# === 4. Gemeinsame Lnge festlegen ===
N = 2_000_000
print(f"Datenstze auf {N} begrenzt.")

primes = primes[:N]
beta_op_vals = beta_op_vals[:N]
beta_kom_vals = beta_kom_vals[:N]
beta_prz_vals = beta_prz_vals[:N]

# === 5. Lambda(n) konstruieren ===
Lambda_np = np.zeros(N, dtype=np.float32)
for p in primes:
    if p <= N:
        Lambda_np[p - 1] = np.log(p)

# === 6. Tensoren erstellen ===
n = torch.arange(1, N + 1, dtype=torch.float32, device=device)
Lambda_n = torch.tensor(Lambda_np, dtype=torch.float32, device=device)
beta_op = torch.tensor(beta_op_vals, dtype=torch.float32, device=device)
beta_kom = torch.tensor(beta_kom_vals, dtype=torch.float32, device=device)
beta_prz = torch.tensor(beta_prz_vals, dtype=torch.float32, device=device)

# === 7. Euler-Freese-Terme ===
terms_op = Lambda_n / torch.pow(n, beta_op)
terms_kom = Lambda_n / torch.pow(n, beta_kom)
terms_prz = Lambda_n / torch.pow(n, beta_prz)
terms_fixed = Lambda_n / torch.pow(n, 0.5)

# === 8. Statistikfunktion ===
def print_stats(name, tensor):
    arr = tensor.cpu().numpy()
    print(f"\nStatistik {name}:")
    print(f"  Mittelwert:    {np.mean(arr):.6f}")
    print(f"  Standardabw.:  {np.std(arr):.6f}")
    print(f"  Minimum:       {np.min(arr):.6f}")
    print(f"  Maximum:       {np.max(arr):.6f}")

print_stats("Operator-Skala", terms_op)
print_stats("Kombi-Skala", terms_kom)
print_stats("Przision-Skala", terms_prz)

# === 9. Plot ===
plt.figure(figsize=(12, 6))
plt.plot(n.cpu(), terms_op.cpu(), alpha=0.4, label='Operator')
plt.plot(n.cpu(), terms_kom.cpu(), alpha=0.4, label='Kombi')
plt.plot(n.cpu(), terms_prz.cpu(), alpha=0.4, label='Przision')
plt.plot(n.cpu(), terms_fixed.cpu(), linestyle='--', alpha=0.7, label='Fix  = 0.5')
plt.legend()
plt.title("Euler-Freese-Terme im Vergleich")
plt.xlabel("n")
plt.ylabel("(n)/n^(n)")
plt.grid(True)
plt.tight_layout()
plt.show()

# === 10. Korrelation mit Zeta-Nullstellen ===
try:
    zeta_zeros = pd.read_csv(zeta_file, header=None)[0].dropna().astype(float).values[:N]
    for name, terms in [("Operator", terms_op), ("Kombi", terms_kom), ("Przision", terms_prz)]:
        arr = terms[:len(zeta_zeros)].cpu().numpy()
        r = pearsonr(arr, zeta_zeros)[0]
        print(f"Korrelation mit Zeta-Nullstellen ({name}): {r:.5f}")
except Exception as e:
    print(f"Korrelation nicht mglich: {e}")
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.stats import pearsonr

# === 1. Gerteeinstellung ===
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Verwende Gert: {device}")

# === 2. Pfade ===
prime_file = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
beta_op_path = '/content/drive/MyDrive/beta_scale_operator_cleaned.csv'
beta_kom_path = '/content/drive/MyDrive/beta_skala_kombi.csv'
beta_prz_path = '/content/drive/MyDrive/beta_skala_przision.csv'
zeta_file = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt'

# === 3. Primzahlen laden und bereinigen ===
prime_data = pd.read_csv(prime_file, header=None)
prime_data = prime_data[pd.to_numeric(prime_data[0], errors='coerce').notna()]
primes = prime_data[0].astype(int).values
N = 2_000_000  # fester Schnitt
primes = primes[:N]
print(f"Datenstze auf {len(primes)} begrenzt.")

# === 4. Beta-Skalen einlesen & bereinigen ===
def load_beta(file_path, col=0):
    df = pd.read_csv(file_path, header=None)
    beta = pd.to_numeric(df[col], errors='coerce').dropna().values
    return beta[:N]

beta_op = load_beta(beta_op_path, col=0)
beta_kom = load_beta(beta_kom_path, col=0)
beta_prz = load_beta(beta_prz_path, col=0)
beta_fix = 0.5 * np.ones(N)

# === 5. Tensoren vorbereiten ===
n = torch.arange(1, N + 1, dtype=torch.float32, device=device)
Lambda_n = torch.zeros(N, dtype=torch.float32, device=device)
for p in primes:
    if p <= N:
        Lambda_n[p - 1] = np.log(p)

def weighted_sum(beta_np):
    beta_tensor = torch.tensor(beta_np, dtype=torch.float32, device=device)
    return Lambda_n / torch.pow(n, beta_tensor)

terms_op = weighted_sum(beta_op)
terms_kom = weighted_sum(beta_kom)
terms_prz = weighted_sum(beta_prz)
terms_fix = weighted_sum(beta_fix)

# === 6. Statistiken ===
def show_stats(name, data):
    data_np = data.cpu().numpy()
    print(f"\nStatistik {name}-Skala:")
    print(f"  Mittelwert:    {data_np.mean():.6f}")
    print(f"  Standardabw.:  {data_np.std():.6f}")
    print(f"  Minimum:       {data_np.min():.6f}")
    print(f"  Maximum:       {data_np.max():.6f}")

show_stats("Operator", terms_op)
show_stats("Kombi", terms_kom)
show_stats("Przision", terms_prz)
show_stats("Fix (0.5)", terms_fix)

# === 7. Korrelation mit Zeta-Nullstellen ===
try:
    zeta_zeros = pd.read_csv(zeta_file, header=None)[0].dropna().astype(float).values[:N]
    for name, terms in [("Operator", terms_op), ("Kombi", terms_kom), ("Przision", terms_prz)]:
        arr = terms[:N].cpu().numpy()
        if np.isfinite(arr).all():
            r = pearsonr(arr, zeta_zeros)[0]
            print(f"Korrelation mit Zeta-Nullstellen ({name}): {r:.5f}")
        else:
            print(f"{name}: Korrelation nicht mglich  NaNs oder Infs im Array")
except Exception as e:
    print("Zeta-Korrelation fehlgeschlagen:", e)

# === 8. Visualisierung ===
plt.figure(figsize=(14, 6))
plt.plot(n.cpu(), terms_op.cpu(), label='Operator', alpha=0.5)
plt.plot(n.cpu(), terms_kom.cpu(), label='Kombi', alpha=0.5)
plt.plot(n.cpu(), terms_prz.cpu(), label='Przision', alpha=0.5)
plt.plot(n.cpu(), terms_fix.cpu(), label='Fix  = 0.5', alpha=0.4, linestyle='--')
plt.legend()
plt.title("Vergleich gewichteter Euler-Freese-Terme")
plt.xlabel("n")
plt.ylabel(r"$\Lambda(n)/n^\beta$")
plt.grid(True)
plt.tight_layout()
plt.show()

# === 9. Fourier-Analyse ===
Y = fft(terms_op.cpu().numpy())
freqs = fftfreq(len(Y))
plt.figure(figsize=(12, 4))
plt.plot(freqs[:len(freqs)//2], np.abs(Y[:len(Y)//2]))
plt.title("Fourier-Spektrum der Operator-basierten Euler-Freese-Terme")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.preprocessing import MinMaxScaler

# --- Konfiguration ---
N = 2_000_000
paths = {
    "Operator Cleaned": "/content/drive/MyDrive/beta_scale_operator_cleaned.csv",
    "Kombi": "/content/drive/MyDrive/beta_skala_kombi.csv",
    "Przision": "/content/drive/MyDrive/beta_skala_praezision_cleaned.csv"
}

# --- Theoretisches Modell: loglog(n) + B2 ---
n = np.arange(2, N + 2)
beta_theo = np.log(np.log(n)) + 0.42278  # B2  0.42278

# --- Ergebnisstruktur ---
results = []

# --- Verarbeitung jeder Beta-Skala ---
for name, path in paths.items():
    try:
        df = pd.read_csv(path, header=None)
        if df.shape[1] == 1:
            beta_vals = df[0].astype(float).values[:N]
        else:
            beta_vals = df[1].astype(float).values[:N]

        # Normalisierung
        scaler = MinMaxScaler()
        beta_norm = scaler.fit_transform(beta_vals.reshape(-1, 1)).flatten()
        theo_norm = scaler.fit_transform(beta_theo.reshape(-1, 1)).flatten()

        # Korrelation und Abweichungen
        r_norm = pearsonr(beta_norm, theo_norm)[0]
        mad = np.mean(np.abs(beta_vals - beta_theo))
        max_dev = np.max(np.abs(beta_vals - beta_theo))

        results.append((name, r_norm, mad, max_dev))
    except Exception as e:
        results.append((name, f"Fehler: {str(e)}", "-", "-"))

# --- Ergebnisse anzeigen ---
results_df = pd.DataFrame(results, columns=["Skala", "Korrelation (normalisiert)", "Mittlere Abw.", "Maximale Abw."])
print(results_df)
```

```python
# --- Bibliotheken ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Parameter ---
beta_path = "/content/drive/MyDrive/beta_skala_kombi_repaired.csv"  # oder andere CSV

# --- Daten laden ---
beta_raw = pd.read_csv(beta_path).iloc[:, -1].astype(float).values
N = len(beta_raw)

# --- Epsilon-Kurve berechnen ---
beta_cumsum = np.cumsum(beta_raw)
epsilon_N = 1 - beta_cumsum

# --- Plot: epsilon(N) ---
plt.figure(figsize=(8, 5))
plt.plot(np.arange(1, N+1), epsilon_N, label=r'$\varepsilon(N)$', linewidth=1)
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.xlabel("N")
plt.ylabel(r"$\varepsilon(N) = 1 - \sum_{n=1}^N \beta(n)$")
plt.title(r"Korrekturterm $\varepsilon(N)$ aus $\beta(n)$")  # <- korrigiert
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Normierung ---
beta_scaled = beta_raw / beta_cumsum[-1]  # Normierung auf Summe = 1
beta_scaled_cumsum = np.cumsum(beta_scaled)
epsilon_scaled = 1 - beta_scaled_cumsum

# --- Vergleichsplot: Original vs. Skaliert ---
plt.figure(figsize=(8, 5))
plt.plot(np.arange(1, N+1), epsilon_N, label="Original", alpha=0.7)
plt.plot(np.arange(1, N+1), epsilon_scaled, label="Normiert", alpha=0.7)
plt.axhline(0, color='black', linestyle='--', linewidth=0.7)
plt.xlabel("N")
plt.ylabel(r"$\varepsilon(N)$")
plt.title(r"Vergleich: $\varepsilon(N)$ vor und nach Normierung")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
