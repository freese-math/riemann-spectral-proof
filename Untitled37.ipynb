```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import ks_2samp
from google.colab import drive

# ‚úÖ Google Drive einbinden
drive.mount('/content/drive')

# ‚úÖ Daten einlesen (Beta-Skala & Zeta-Nullstellen)
beta_theta_path = "/content/drive/MyDrive/Beta-Theta-Operator-Analyse.csv"
zeta_nullstellen_path = "/content/drive/MyDrive/zeros6.txt"

beta_df = pd.read_csv(beta_theta_path)
zeta_df = pd.read_csv(zeta_nullstellen_path)

# ‚úÖ Daten extrahieren
n_values = beta_df["N"].values
beta_values = beta_df["Beta-Skala ohne Theta"].values
eigenvalues = beta_df["Gemessene Eigenwerte"].values
zeta_nullstellen = zeta_df["Imagin√§rteil"].values  # Zeta-Nullstellen

# ‚úÖ 1Ô∏è‚É£ Vergleich Beta-Skala vs. Zeta-Nullstellen
plt.figure(figsize=(8,5))
plt.scatter(n_values, beta_values, color='blue', label="Beta-Werte")
plt.scatter(n_values[:len(zeta_nullstellen)], zeta_nullstellen, color='red', alpha=0.5, label="Zeta-Nullstellen")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Beta-Skala vs. Zeta-Nullstellen")
plt.show()

# ‚úÖ 2Ô∏è‚É£ Korrelation Beta-Skala vs. Zeta-Nullstellen
correlation = np.corrcoef(beta_values[:len(zeta_nullstellen)], zeta_nullstellen)[0,1]
print(f"üîπ Korrelation Beta-Skala vs. Zeta-Nullstellen: {correlation:.5f}")

# ‚úÖ 3Ô∏è‚É£ Test: Spektrale Strukturen (GOE/GUE vs. Beta)
level_spacings_beta = np.diff(np.sort(beta_values))
level_spacings_zeta = np.diff(np.sort(zeta_nullstellen))

ks_stat, ks_p_value = ks_2samp(level_spacings_beta, level_spacings_zeta)
print(f"üîπ Kolmogorov-Smirnov-Test (Beta vs. Zeta-Nullstellen): p-Wert = {ks_p_value:.5f}")

# üìà **Histogramm Level-Abst√§nde**
plt.figure(figsize=(8,5))
plt.hist(level_spacings_beta, bins=50, alpha=0.5, label="Beta-Werte", density=True)
plt.hist(level_spacings_zeta, bins=50, alpha=0.5, label="Zeta-Nullstellen", density=True)
plt.xlabel("Level-Spacings")
plt.ylabel("Dichte")
plt.legend()
plt.title("Level-Abst√§nde: Beta-Skala vs. Zeta-Nullstellen")
plt.show()

# ‚úÖ 4Ô∏è‚É£ Beta-Skala als Schr√∂dinger-Operator
def beta_schroedinger(n, A, B, C):
    return A * n**B + C * np.sin(n)

popt_schroedinger, _ = curve_fit(beta_schroedinger, n_values, beta_values)
A_schr, B_schr, C_schr = popt_schroedinger

plt.figure(figsize=(8,5))
plt.scatter(n_values, beta_values, color='purple', label="Beta-Werte")
plt.plot(n_values, beta_schroedinger(n_values, *popt_schroedinger), 'r--', label="Schr√∂dinger-Approximation")
plt.legend()
plt.xlabel("Index")
plt.ylabel("Beta-Wert")
plt.title("Beta-Skala als Schr√∂dinger-Wellenfunktion")
plt.show()

# ‚úÖ 5Ô∏è‚É£ Schwarzschild-Kr√ºmmung f√ºr Beta-Skala testen
def schwarzschild_kr√ºmmung(n, M):
    return 1 - 2 * M / n

popt_schwarzschild, _ = curve_fit(schwarzschild_kr√ºmmung, n_values, beta_values)
M_schwarzschild = popt_schwarzschild[0]

plt.figure(figsize=(8,5))
plt.scatter(n_values, beta_values, color='green', label="Beta-Werte")
plt.plot(n_values, schwarzschild_kr√ºmmung(n_values, M_schwarzschild), 'r--', label="Schwarzschild-Kr√ºmmung")
plt.legend()
plt.xlabel("Index")
plt.ylabel("Beta-Wert")
plt.title("Beta-Skala vs. Schwarzschild-Kr√ºmmung")
plt.show()

# ‚úÖ 6Ô∏è‚É£ **Beta-Skala gegen Operatoren der Zufallsmatrizen testen**
level_spacings_beta = np.diff(np.sort(beta_values))
level_spacings_goe = np.random.normal(0, 1, size=len(level_spacings_beta))

ks_stat_goe, ks_p_value_goe = ks_2samp(level_spacings_beta, level_spacings_goe)
print(f"üîπ KS-Test Beta vs. GOE-Zufallsmatrix: p-Wert = {ks_p_value_goe:.5f}")

# üìä **Ergebnisse ausgeben**
print(f"üìä Ergebnisse der RH-Pr√ºfung mit Beta-Skala:")
print(f"üîπ Korrelation Beta-Skala vs. Zeta-Nullstellen: {correlation:.5f}")
print(f"üîπ KS-Test f√ºr Level-Spacings Beta vs. Zeta: p-Wert = {ks_p_value:.5f}")
print(f"üîπ KS-Test Beta vs. GOE: p-Wert = {ks_p_value_goe:.5f}")
print(f"üîπ Beta-Skala als Schr√∂dinger-Gleichung: A={A_schr:.4f}, B={B_schr:.4f}, C={C_schr:.4f}")
print(f"üîπ Schwarzschild-Kr√ºmmung f√ºr Beta-Skala: M={M_schwarzschild:.4f}")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from scipy.optimize import curve_fit
from scipy.stats import ks_2samp
from google.colab import drive

# ‚úÖ Google Drive einbinden
drive.mount('/content/drive')

# ‚úÖ Nullstellen einlesen (Odlyzko)
zeta_nullstellen_path = "/content/drive/MyDrive/zeros6.txt"
zeta_nullstellen = np.loadtxt(zeta_nullstellen_path)[:2000000]  # Nur 2 Mio Nullstellen nutzen

# ‚úÖ Beta-Werte einlesen
beta_theta_path = "/content/drive/MyDrive/Beta-Theta-Operator-Analyse.csv"
beta_df = pd.read_csv(beta_theta_path)
n_values = beta_df["N"].values[:2000000]  # Nur 2 Mio Werte
beta_values = beta_df["Beta-Skala ohne Theta"].values[:2000000]

# ‚úÖ 1Ô∏è‚É£ Selbstadjungiertheit von H pr√ºfen
def check_self_adjoint(H_matrix):
    return np.allclose(H_matrix, H_matrix.T)

# Beispiel f√ºr eine Hamilton-Matrix (diskrete Version von H)
H_matrix = np.diag(beta_values) + np.diag(beta_values[:-1], k=1) + np.diag(beta_values[:-1], k=-1)

# Test auf Selbstadjungiertheit
is_self_adjoint = check_self_adjoint(H_matrix)
print(f"‚úÖ Ist H selbstadjungiert? {'Ja' if is_self_adjoint else 'Nein'}")

# ‚úÖ 2Ô∏è‚É£ Eigenwerte von H berechnen
eigenvalues, eigenvectors = eigh(H_matrix)

# ‚úÖ 3Ô∏è‚É£ Vergleich der Eigenwerte mit Zeta-Nullstellen
plt.figure(figsize=(8,5))
plt.plot(eigenvalues[:50], 'bo-', label="H-Eigenwerte")
plt.plot(zeta_nullstellen[:50], 'ro-', alpha=0.5, label="Zeta-Nullstellen")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.title("Vergleich: H-Eigenwerte vs. Zeta-Nullstellen")
plt.legend()
plt.show()

# ‚úÖ 4Ô∏è‚É£ Kolmogorov-Smirnov-Test (H vs. Zeta-Nullstellen)
ks_stat, ks_p_value = ks_2samp(eigenvalues[:2000], zeta_nullstellen[:2000])
print(f"üîπ KS-Test (H-Eigenwerte vs. Zeta-Nullstellen): p-Wert = {ks_p_value:.5f}")

# ‚úÖ 5Ô∏è‚É£ Optimierter Hamilton-Operator anpassen (Kurvenfit)
def beta_operator(n, A, B, C):
    return A * n**B + C * np.sin(n)

popt_H, _ = curve_fit(beta_operator, n_values, beta_values)
A_H, B_H, C_H = popt_H
print(f"üìä Optimierte H-Parameter: A={A_H:.4f}, B={B_H:.4f}, C={C_H:.4f}")

# ‚úÖ 6Ô∏è‚É£ Darstellung der optimierten Beta-Skala als Operator-Funktion
plt.figure(figsize=(8,5))
plt.scatter(n_values, beta_values, color='blue', alpha=0.5, label="Beta-Werte")
plt.plot(n_values, beta_operator(n_values, *popt_H), 'r--', label="Optimierter H-Operator")
plt.xlabel("N")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Optimierte Beta-Skala als Operator-Funktion")
plt.show()

# ‚úÖ Ergebnisse zusammenfassen
print("üìä Ergebnisse der RH-Pr√ºfung mit H-Operator:")
print(f"‚úÖ Ist H selbstadjungiert? {'Ja' if is_self_adjoint else 'Nein'}")
print(f"üîπ KS-Test f√ºr H-Eigenwerte vs. Zeta-Nullstellen: p-Wert = {ks_p_value:.5f}")
print(f"üîπ Optimierte H-Parameter: A={A_H:.4f}, B={B_H:.4f}, C={C_H:.4f}")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy.optimize import curve_fit
from google.colab import drive

# ‚úÖ Google Drive einbinden
drive.mount('/content/drive')

# ‚úÖ Daten einlesen
beta_theta_path = "/content/drive/MyDrive/Beta-Theta-Operator-Analyse.csv"
zeta_nullstellen_path = "/content/drive/MyDrive/zeros6.txt"

beta_df = pd.read_csv(beta_theta_path)
zeta_nullstellen = np.loadtxt(zeta_nullstellen_path)[:1000]  # Begrenzen auf 1000 f√ºr Performance

# ‚úÖ Daten extrahieren
n_values = beta_df["N"].values[:1000]  # Begrenzung f√ºr Rechenzeit
beta_values = beta_df["Beta-Skala ohne Theta"].values[:1000]

# ‚úÖ 1Ô∏è‚É£ Definiere das Potential V(n) der Beta-Skala
def V(n, A, B, C, D, E, F, Phi):
    return A * n**B + C * np.log(n) + D * n**-1 + E * np.sin(F * n + Phi)

# ‚úÖ 2Ô∏è‚É£ Fit der Beta-Skala auf das Potential
popt, _ = curve_fit(V, n_values, beta_values, p0=[1.8, 0.64, 2488, 0.9, 0.0007, 1.06, -40])

# ‚úÖ 3Ô∏è‚É£ Hamilton-Matrix f√ºr Schr√∂dinger-Gleichung aufbauen
N = len(n_values)
diagonal = V(n_values, *popt)
off_diagonal = -0.5 * np.ones(N-1)  # Tridiagonale N√§herung

# ‚úÖ 4Ô∏è‚É£ Eigenwerte des Operators berechnen
eigenvalues, _ = eigh_tridiagonal(diagonal, off_diagonal)

# ‚úÖ 5Ô∏è‚É£ Vergleich der Eigenwerte mit Beta-Werten und Zeta-Nullstellen
plt.figure(figsize=(10,6))
plt.scatter(n_values, eigenvalues, color='blue', label="H-Eigenwerte")
plt.scatter(n_values, beta_values, color='red', alpha=0.5, label="Beta-Werte")
plt.scatter(n_values[:len(zeta_nullstellen)], zeta_nullstellen, color='green', alpha=0.5, label="Zeta-Nullstellen")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Vergleich: H-Eigenwerte vs. Beta-Skala vs. Zeta-Nullstellen")
plt.show()

# ‚úÖ 6Ô∏è‚É£ Korrelationen berechnen
correlation_beta = np.corrcoef(eigenvalues, beta_values)[0,1]
correlation_zeta = np.corrcoef(eigenvalues[:len(zeta_nullstellen)], zeta_nullstellen)[0,1]

print(f"üìä Ergebnisse der RH-Pr√ºfung mit H-Operator:")
print(f"üîπ Korrelation H-Eigenwerte vs. Beta-Skala: {correlation_beta:.5f}")
print(f"üîπ Korrelation H-Eigenwerte vs. Zeta-Nullstellen: {correlation_zeta:.5f}")

# ‚úÖ 7Ô∏è‚É£ Test auf Selbstadjungiertheit
eigvals_diff = np.abs(np.imag(eigenvalues))
if np.max(eigvals_diff) < 1e-10:
    print("‚úÖ H ist selbstadjungiert!")
else:
    print("‚ùå H ist nicht selbstadjungiert!")
```

```python
# ‚úÖ Bibliotheken importieren
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# ‚úÖ Zielfunktion definieren: Wir wollen H e^(Œ≤ œÄ i) + 1 ‚âà 0 l√∂sen
def equation_to_optimize(beta, H):
    result = H * np.exp(beta * np.pi * 1j) + 1
    return np.abs(result)  # Minimieren die absolute Abweichung von 0

# ‚úÖ Startwerte f√ºr die Optimierung
beta_initial = 0.914  # Startwert f√ºr Œ≤
H_initial = 1.0       # Annahme: H = 1, falls nicht, k√∂nnen wir auch H optimieren

# ‚úÖ Optimierung durchf√ºhren
opt_result = opt.minimize_scalar(equation_to_optimize, args=(H_initial), method="brent")

# ‚úÖ Optimiertes Œ≤ ausgeben
beta_optimized = opt_result.x
print(f"üîπ Optimiertes Œ≤: {beta_optimized:.10f}")

# ‚úÖ Berechnung des optimierten Funktionswerts
final_value = equation_to_optimize(beta_optimized, H_initial)
print(f"üîπ Wert von |H e^(Œ≤ œÄ i) + 1| f√ºr optimiertes Œ≤: {final_value:.10e}")

# ‚úÖ Visualisierung: Plot der Funktion √ºber einen Bereich von Œ≤-Werten
beta_values = np.linspace(0.8, 1.0, 200)  # Werte von Œ≤ in der N√§he des Startwerts
errors = [equation_to_optimize(beta, H_initial) for beta in beta_values]

plt.figure(figsize=(8,5))
plt.plot(beta_values, errors, 'b-', label="Fehlermetrik |H e^(Œ≤ œÄ i) + 1|")
plt.axvline(beta_optimized, color='r', linestyle='--', label=f"Optimiertes Œ≤ = {beta_optimized:.10f}")
plt.xlabel("Œ≤-Wert")
plt.ylabel("Fehler |H e^(Œ≤ œÄ i) + 1|")
plt.legend()
plt.title("Optimierung von Œ≤ f√ºr H e^(Œ≤ œÄ i) + 1 = 0")
plt.show()
```

```python
# ‚úÖ Ben√∂tigte Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.special import jtheta

# ‚úÖ Jacobi-Theta-Korrektur f√ºr Beta-Skala definieren
def beta_mit_theta(n, A, B, C, a, c):
    """
    Modell zur nichtlinearen Approximation der Beta-Skala mit der Jacobi-Theta-Funktion.
    """
    theta_val = np.array([jtheta(3, c * np.pi * val, np.exp(-np.pi)) for val in n])
    return A * n**B + C + a * np.exp(-c * n) * theta_val

# ‚úÖ Simulierte Beta-Skala-Werte (Beispieldaten)
n_values = np.linspace(0.01, 10, 300)  # Werte f√ºr n
true_params = [0.5, 0.1, 0.01, 0.131415, 0.006738]  # Wahre Parameter (gesch√§tzt)
beta_values = beta_mit_theta(n_values, *true_params) + np.random.normal(0, 0.001, len(n_values))  # Mit Rauschen

# ‚úÖ Anfangswerte f√ºr Curve-Fitting
p0 = [0.4, 0.05, 0.001, 0.1, 0.005]

# ‚úÖ Curve Fit f√ºr Theta-Korrektur durchf√ºhren
try:
    popt, _ = curve_fit(beta_mit_theta, n_values, beta_values, p0=p0, maxfev=10000)
    A_opt, B_opt, C_opt, a_opt, c_opt = popt
    print(f"üîπ Optimierte Theta-Parameter: A={A_opt:.6f}, B={B_opt:.6f}, C={C_opt:.6f}, a={a_opt:.6f}, c={c_opt:.6f}")
except RuntimeError:
    print("‚ö†Ô∏è Curve_fit fehlgeschlagen! Alternative Methode erforderlich.")

# ‚úÖ Modell mit optimierten Parametern plotten
plt.figure(figsize=(8, 5))
plt.scatter(n_values, beta_values, s=5, label="Original-Daten (mit Rauschen)", color='blue', alpha=0.5)
plt.plot(n_values, beta_mit_theta(n_values, *popt), color='red', label="Gefittetes Modell")
plt.xlabel("n-Werte")
plt.ylabel("Œ≤(n)")
plt.legend()
plt.title("Nichtlineare Anpassung der Beta-Skala mit Jacobi-Theta-Korrektur")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import pearsonr
import seaborn as sns

# üîπ Dateien hochladen (manuell in Colab, falls nicht automatisch geladen)
from google.colab import files

uploaded = files.upload()

# üîπ Daten laden
freese_function = pd.read_csv("freese_function_odlyzko_cleaned.csv")  # Anpassung des Dateinamens
optimized_results = pd.read_csv("optimized_freese_function_results.csv")
fit_data = pd.read_csv("freese_function_fit.csv")
fractal_data = pd.read_csv("fractal_dimension_results.csv")

# üîπ Nullstellen der Zeta-Funktion aus Datei laden (Anpassung des Dateinamens)
nullstellen = np.loadtxt("zeros6.txt")

# üîπ Log-Log-Regression f√ºr Fraktale Dimension
x = fractal_data.iloc[:, 1].values  # log(1/epsilon)
y = fractal_data.iloc[:, 2].values  # log(N(epsilon))

# üîπ Lineare Regression zur Bestimmung der Fraktaldimension
def linear_model(x, m, b):
    return m * x + b

popt, _ = curve_fit(linear_model, x, y)
fractal_dimension = popt[0]

print(f"üîπ Berechnete Fraktaldimension: {fractal_dimension:.4f}")

# üîπ Vergleich der Freese-Funktion mit echten Nullstellen
plt.figure(figsize=(10, 5))
plt.plot(nullstellen, np.zeros_like(nullstellen), 'ro', markersize=3, label="Echte Nullstellen")
plt.plot(freese_function['x'], freese_function['y'], 'b-', alpha=0.6, label="Freese-Funktion")
plt.xlabel("Re(s)")
plt.ylabel("Wert")
plt.title("Vergleich der Freese-Funktion mit echten Nullstellen")
plt.legend()
plt.show()

# üîπ Statistische Korrelation Freese vs. echte Nullstellen
corr, _ = pearsonr(nullstellen[:len(freese_function)], freese_function['y'])
print(f"üîπ Korrelation zwischen Freese-Funktion und echten Nullstellen: {corr:.4f}")

# üîπ Visualisierung der Abweichungen
abweichung = np.abs(nullstellen[:len(freese_function)] - freese_function['y'])

plt.figure(figsize=(8, 5))
sns.histplot(abweichung, bins=50, kde=True)
plt.xlabel("Abweichung")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm der Abweichungen")
plt.show()

# üîπ Hypothese testen: Ist die Freese-Funktion unabh√§ngig? (FFF independent)
# Wir betrachten die Autokorrelation der Residuen (wenn unabh√§ngig, sollte sie gering sein)
autokorrelation = np.corrcoef(abweichung[:-1], abweichung[1:])[0, 1]
print(f"üîπ Autokorrelation der Abweichungen: {autokorrelation:.4f}")

# Hypothese: Falls Autokorrelation ‚âà 0, ist FFF unabh√§ngig
if abs(autokorrelation) < 0.1:
    print("‚úÖ Die Hypothese, dass FFF unabh√§ngig ist, wird gest√ºtzt!")
else:
    print("‚ùå Die Hypothese wird geschw√§cht ‚Äì es gibt eine signifikante Korrelation.")

# üîπ Fraktale Dimension visualisieren
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label="Datenpunkte")
plt.plot(x, linear_model(x, *popt), 'r--', label=f"Fit: D = {fractal_dimension:.4f}")
plt.xlabel("log(1/epsilon)")
plt.ylabel("log(N(epsilon))")
plt.title("Fraktale Dimension durch Log-Log Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üîπ Hardcodierte Dateipfade (Google Drive)
nullstellen_path = "/content/drive/MyDrive/zeros6.txt"
freese_cleaned_path = "/content/drive/MyDrive/freese_function_odlyzko_cleaned.csv"
freese_optimized_path = "/content/drive/MyDrive/optimized_freese_function_results.csv"
freese_fit_path = "/content/drive/MyDrive/freese_function_fit.csv"
fractal_dimension_path = "/content/drive/MyDrive/fractal_dimension_results.csv"

# üîπ 1Ô∏è‚É£ Nullstellen einlesen
nullstellen = np.loadtxt(nullstellen_path)
print(f"üìå {len(nullstellen)} Nullstellen geladen.")

# üîπ 2Ô∏è‚É£ Freese-Funktion-Daten einlesen
freese_cleaned = pd.read_csv(freese_cleaned_path)
freese_optimized = pd.read_csv(freese_optimized_path)
freese_fit = pd.read_csv(freese_fit_path)

# üîπ 3Ô∏è‚É£ Vergleich der Nullstellen mit Freese-Werten
plt.figure(figsize=(10, 5))
plt.plot(nullstellen, label="Echte Nullstellen", marker='o', linestyle='', alpha=0.5)
plt.plot(freese_cleaned['x'], freese_cleaned['y'], label="Freese (Cleaned)", linestyle='--')
plt.plot(freese_optimized['x'], freese_optimized['y'], label="Freese (Optimized)", linestyle='-.')
plt.plot(freese_fit['x'], freese_fit['y'], label="Freese (Fit)", linestyle=':')
plt.xlabel("Index")
plt.ylabel("Wert")
plt.title("Vergleich der Nullstellen mit der Freese-Funktion")
plt.legend()
plt.grid()
plt.show()

# üîπ 4Ô∏è‚É£ Fehleranalyse (Differenz zwischen Freese & Nullstellen)
freese_errors = np.abs(freese_cleaned['y'] - nullstellen[:len(freese_cleaned)])

plt.figure(figsize=(10, 4))
plt.plot(freese_errors, label="Abweichung Freese vs. Nullstellen")
plt.xlabel("Index")
plt.ylabel("Fehler")
plt.title("Fehleranalyse: Freese-Funktion vs. echte Nullstellen")
plt.legend()
plt.grid()
plt.show()

# üîπ 5Ô∏è‚É£ Fraktaldimension auswerten
fractal_data = pd.read_csv(fractal_dimension_path)
x = fractal_data.iloc[:, 0].values  # log(1/epsilon)
y = fractal_data.iloc[:, 1].values  # log(N(epsilon))

# üîπ Lineare Regression f√ºr Fraktaldimension
coeffs = np.polyfit(x, y, 1)
fractal_dimension = coeffs[0]

print(f"üîπ Berechnete Fraktaldimension: {fractal_dimension:.4f}")

# üîπ 6Ô∏è‚É£ Visualisierung der Fraktaldimension
plt.figure(figsize=(8, 5))
plt.scatter(x, y, label="Datenpunkte", alpha=0.6)
plt.plot(x, np.polyval(coeffs, x), 'r--', label=f"Fit (D={fractal_dimension:.4f})")
plt.xlabel("log(1/epsilon)")
plt.ylabel("log(N(epsilon))")
plt.title("Fraktaldimension Analyse")
plt.legend()
plt.grid()
plt.show()
```
