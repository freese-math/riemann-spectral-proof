```python
# === Google Drive Mount (für Colab) ===
from google.colab import drive
drive.mount('/content/drive')

# === Bibliotheken laden ===
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow_probability as tfp
import matplotlib.pyplot as plt

# === 1️⃣ Zeta-Nullstellen laden (erste 1000 für Basis) ===
zeta_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.csv"
zeta_df = pd.read_csv(zeta_path, header=None, skiprows=1)
zeta_np = zeta_df.iloc[:, 0].astype(float).values[:1000]
gamma = tf.constant(zeta_np, dtype=tf.float32)

# === 2️⃣ θ-Parameter initialisieren ===
theta = tf.Variable(1.0, dtype=tf.float32)

# === 3️⃣ H-Operator bauen: H(θ) = diag(gamma) + θ * (Off-Diagonale) ===
def build_H(gamma, theta):
    N = tf.shape(gamma)[0]
    diag = tf.linalg.diag(gamma)
    off_diag = tf.linalg.diag(tf.ones(N-1), k=1) + tf.linalg.diag(tf.ones(N-1), k=-1)
    return diag + theta * off_diag

# === 4️⃣ Verlustfunktion: 1 - Korrelation der Eigenwerte mit Nullstellen ===
@tf.function
def compute_loss(theta, gamma):
    H = build_H(gamma, theta)
    eigvals = tf.linalg.eigvalsh(H)
    corr = tfp.stats.correlation(tf.expand_dims(gamma, axis=0),
                                 tf.expand_dims(eigvals, axis=0),
                                 sample_axis=1)[0, 0]
    return 1.0 - corr

# === 5️⃣ Optimierung ===
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss_log = []

for step in range(200):
    with tf.GradientTape() as tape:
        loss = compute_loss(theta, gamma)
    grads = tape.gradient(loss, [theta])
    opt.apply_gradients(zip(grads, [theta]))
    loss_log.append(loss.numpy())
    if step % 20 == 0:
        print(f"Schritt {step:03d} | Loss: {loss:.6f} | θ: {theta.numpy():.6f}")

# === 6️⃣ Visualisierung der Optimierung ===
plt.plot(loss_log, label="Loss")
plt.title("Optimierungsverlauf: Loss vs. Schritt")
plt.xlabel("Schritt")
plt.ylabel("Loss (1 - Korrelation)")
plt.grid()
plt.legend()
plt.show()

# === 7️⃣ Ergebnisplot: Nullstellen vs. Eigenwerte von H(θ) ===
H_opt = build_H(gamma, theta)
eigvals_final = tf.linalg.eigvalsh(H_opt).numpy()

plt.figure(figsize=(6, 4))
plt.plot(gamma.numpy(), label="Zeta-Nullstellen", linewidth=1)
plt.plot(eigvals_final, label="Eigenwerte von H(θ)", linestyle="--", linewidth=1)
plt.legend()
plt.grid()
plt.title("Vergleich: Nullstellen vs. optimierte Eigenwerte")
plt.show()

print(f"✅ Optimierter θ-Wert: {theta.numpy():.6f}")
```

```python
# ✅ Bibliotheken importieren
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# === 1️⃣ Daten einlesen (v5 enthält Zeta-Nullstellen und Beta-Spalte)
pfad = "/content/drive/MyDrive/beta_skala_v5.csv"
df = pd.read_csv(pfad, header=None)

# Ziel: Zeta-Nullstellen (Spalte 0), Start: Beta v5 (Spalte 1)
gamma = tf.constant(df.iloc[:1000, 0].astype(float).values.reshape(1, -1), dtype=tf.float32)
theta = tf.Variable(df.iloc[:1000, 1].astype(float).values.reshape(1, -1), dtype=tf.float32)

# === 2️⃣ Optimierer
opt = tf.keras.optimizers.Adam(learning_rate=0.05)
loss_history = []

# === 3️⃣ Verlustfunktion (MSE + Korrelation)
def compute_loss(theta, gamma):
    mse = tf.reduce_mean(tf.square(theta - gamma))
    gamma_centered = gamma - tf.reduce_mean(gamma)
    theta_centered = theta - tf.reduce_mean(theta)
    corr_num = tf.reduce_sum(gamma_centered * theta_centered)
    corr_den = tf.sqrt(tf.reduce_sum(tf.square(gamma_centered)) * tf.reduce_sum(tf.square(theta_centered)))
    correlation = corr_num / corr_den
    return mse + (1 - correlation)

# === 4️⃣ Trainingsschleife
for step in range(200):
    with tf.GradientTape() as tape:
        loss = compute_loss(theta, gamma)
    grads = tape.gradient(loss, [theta])
    opt.apply_gradients(zip(grads, [theta]))
    loss_history.append(loss.numpy())
    if step % 20 == 0:
        print(f"Step {step:3d} | Loss: {loss.numpy():.6f}")

# === 5️⃣ Plot: Loss-Verlauf
plt.figure(figsize=(9, 4))
plt.plot(loss_history)
plt.title("Verlauf der Loss-Funktion")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# === 6️⃣ Plot: Ziel vs. Optimiertes Ergebnis
plt.figure(figsize=(9, 4))
plt.plot(gamma.numpy().flatten(), label="Ziel (Zeta-Nullstellen)")
plt.plot(theta.numpy().flatten(), label="Optimiert (Beta)")
plt.title("Vergleich: Ziel vs. Optimiertes Ergebnis")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# === Dateien definieren ===
pfad_odlyzko = "/content/drive/MyDrive/zeros1.txt"
pfad_lmfdb = "/content/drive/MyDrive/lmfd_zeros_100k.txt"
pfad_zeros6 = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt"
pfad_beta = "/content/drive/MyDrive/beta_skala_v5_biasfrei_korrigiert.csv"

# === Robustere Ladefunktionen ===
def lade_txt_liste(pfad):
    with open(pfad, "r") as f:
        werte = [line.strip() for line in f if line.strip()]
    return np.array([float(w) for w in werte if not any(c.isalpha() for c in w)])

def lade_lmfdb(pfad):
    with open(pfad, "r") as f:
        werte = []
        for zeile in f:
            teile = zeile.strip().split()
            if len(teile) >= 2:
                try:
                    werte.append(float(teile[1]))
                except ValueError:
                    continue
    return np.array(werte)

def lade_beta(pfad):
    df = pd.read_csv(pfad, header=0)  # Header überspringen
    if df.shape[0] == 2:
        df = df.T
    return df.iloc[:, 0].astype(float).values

# === Lade alle Daten
odlyzko = lade_txt_liste(pfad_odlyzko)
lmfdb = lade_lmfdb(pfad_lmfdb)
zeros6 = lade_txt_liste(pfad_zeros6)
beta = lade_beta(pfad_beta)

# === Vergleichsfunktion
def vergleich(a, b, label_a, label_b):
    n = min(len(a), len(b))
    if n == 0:
        print(f"❌ {label_a} vs. {label_b}: keine Vergleichsdaten!")
        return
    diff = np.abs(a[:n] - b[:n])
    print(f"🔍 {label_a} vs. {label_b}")
    print(f"   ✅ Mittelwert Fehler: {np.mean(diff):.10f}")
    print(f"   ✅ Standardabweichung: {np.std(diff):.10f}")

# === Vergleiche ausgeben
vergleich(odlyzko, lmfdb, "Odlyzko", "LMFDB")
vergleich(odlyzko, beta, "Odlyzko", "Beta v5 biasfrei")
vergleich(lmfdb, beta, "LMFDB", "Beta v5 biasfrei")
vergleich(zeros6, beta, "zeros6", "Beta v5 biasfrei")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameterbereich
N = np.linspace(1, 1000, 1000)

# Beispiel-Frequenzen und Amplituden (ähnlich FFF-Spektrum)
k_vals = np.arange(1, 21)  # 20 Frequenzen
frequencies = np.log(k_vals + 1)  # Logarithmisch verteilt (wie Beta-Struktur)
amplitudes = 1 / k_vals  # Abfallende Amplituden (Fourier-typisch)

# 1. Fourier-Summe F(N) erzeugen
F_N = np.zeros_like(N, dtype=complex)
for A_k, f_k in zip(amplitudes, frequencies):
    F_N += A_k * np.exp(2j * np.pi * f_k * N)

# 2. Zeitbereichsmodell Θ(t): explizite quasiperiodische Struktur
t = N
A, beta = 1.2, 0.48
C, D, E = -0.3, 5, 0.8
w, phi = 2 * np.pi * 0.1, np.pi / 4
Theta_t = A * t**beta + C * np.log(t) + D / t + E * np.sin(w * np.log(t) + phi)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(N, F_N.real, label='FFF-Summe (Re Teil)', color='blue')
plt.plot(t, Theta_t, label='Explizites Modell Θ(t)', linestyle='--', color='orange')
plt.title("Vergleich: FFF-Fourier-Summe vs. explizite Theta(t)-Form")
plt.xlabel("N")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Installiere Widgets (nur in Colab nötig)
!pip install -q ipywidgets

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import matplotlib as mpl

# Visualeinstellungen
mpl.rcParams['figure.figsize'] = (10, 5)
mpl.rcParams['axes.grid'] = True

# Definition der expliziten Theta(t) – Form
def theta_function(t, A, beta, C, D, E, w, phi):
    return A * t**beta + C * np.log(t) + D * t**(-1) + E * np.sin(w * np.log(t) + phi)

# Definition der FFF als Fourier-Reihe mit rekursiver Frequenzstruktur
def fff_function(N, coeffs, freqs):
    summe = np.zeros_like(N, dtype=np.complex128)
    for a, f in zip(coeffs, freqs):
        summe += a * np.exp(2j * np.pi * f * N)
    return summe.real

# Interaktive Darstellung
@interact(
    A=FloatSlider(value=1.0, min=0, max=3, step=0.1),
    beta=FloatSlider(value=0.48, min=0.3, max=0.6, step=0.01),
    C=FloatSlider(value=0.5, min=-1, max=1, step=0.1),
    D=FloatSlider(value=0.1, min=-1, max=1, step=0.1),
    E=FloatSlider(value=0.5, min=0, max=2, step=0.1),
    w=FloatSlider(value=2.5, min=0.1, max=10, step=0.1),
    phi=FloatSlider(value=0, min=0, max=2*np.pi, step=0.1)
)
def plot_theta_vs_fff(A, beta, C, D, E, w, phi):
    t = np.linspace(1, 1000, 1000)
    theta = theta_function(t, A, beta, C, D, E, w, phi)

    # Beispielhafte rekursive Frequenzstruktur (Fibonacci-artig)
    k = np.arange(1, 10)
    coeffs = 1 / (k ** 1.5)
    freqs = 0.000005 * k**1.618  # Goldene Potenzierung

    fff = fff_function(t, coeffs, freqs)

    plt.figure(figsize=(12, 5))
    plt.plot(t, theta, label=r'$\Theta(t)$ (Siegel-Theta-Form)', color='blue')
    plt.plot(t, fff, label='FFF (rekursive Fourierstruktur)', color='orange', linestyle='--')
    plt.title("Vergleich: Explizite Theta-Darstellung vs. FFF-Fourier-Form")
    plt.xlabel("t (bzw. N)")
    plt.ylabel("Funktionswert")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# --- Symbolische Definition (SymPy) ---
from sympy import symbols, sin, log, Sum, pi, oo, lambdify, summation
import numpy as np
import matplotlib.pyplot as plt

# Symbole definieren
N, s = symbols('N s', real=True)
A, B, C, beta, omega, phi = symbols('A B C beta omega phi', real=True)

# Freese-Funktion L(N)
L_N = A * N**beta + C * log(N) + B * sin(omega * N + phi)

# Freese-Zetafunktion: ζ_F(s)
zeta_F = Sum(1 / L_N**s, (N, 1, oo))

# --- Numerische Implementierung ---
# Parameter: Angepasst aus deiner notariellen Version (28.2.2025)
param = {
    A: 1.882795,
    B: 1.0,
    C: 2488.144455,
    beta: 0.916977,
    omega: 2 * pi,
    phi: 0.0
}

# L(N) numerisch auswerten
L_func = lambdify(N, L_N.subs(param), modules="numpy")

# ζ_F(s) numerisch berechnen
def freese_zeta_numeric(s_val, N_max=10000):
    N_vals = np.arange(1, N_max + 1)
    L_vals = L_func(N_vals)
    return np.sum(1 / L_vals**s_val)

# Beispielwert
z_val = freese_zeta_numeric(2)
print(f"ζ_F(2) ≈ {z_val:.10f}")

# --- Plot von ζ_F(s) ---
s_range = np.linspace(1.1, 5, 400)
zeta_vals = [freese_zeta_numeric(s_val, N_max=3000) for s_val in s_range]

plt.figure(figsize=(8, 5))
plt.plot(s_range, zeta_vals, label=r'$\zeta_F(s)$', color='darkblue')
plt.xlabel("s")
plt.ylabel(r"$\zeta_F(s)$")
plt.title("Freese-Zetafunktion ζ_F(s)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import mpmath as mp

# Setze Dezimalstellen (z.B. 50)
mp.mp.dps = 50

# Parameterwerte aus deiner FF-Formel
A = mp.mpf("1.882795")
B = mp.mpf("1.0")
C = mp.mpf("2488.144455")
beta = mp.mpf("0.916977")
omega = 2 * mp.pi
phi = mp.mpf("0.0")

# L(N)-Funktion
def L(N):
    N = mp.mpf(N)
    return A * N**beta + C * mp.log(N) + B * mp.sin(omega * N + phi)

# Freese-Zetafunktion ζ_F(s)
def freese_zeta_mpmath(s, N_max=10000):
    s = mp.mpf(s)
    total = mp.mpf(0)
    for N in range(1, N_max + 1):
        term = 1 / L(N)**s
        total += term
    return total

# Beispiel: ζ_F(2) mit hoher Genauigkeit
zeta_f_val = freese_zeta_mpmath(2, N_max=50000)
print("ζ_F(2) ≈", zeta_f_val)
```

```python
# --- Colab-ready Setup ---
import numpy as np
import matplotlib.pyplot as plt
import torch

# CUDA aktivieren
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Verwendetes Gerät:", device)

# --- Definition des Operators H_beta ---
def construct_operator_matrix(N, beta):
    """Diskretisierter Operator H_beta als Matrix (GPU-optimiert)"""
    i = torch.arange(1, N+1, device=device).float()
    j = torch.arange(1, N+1, device=device).float()
    ii, jj = torch.meshgrid(i, j, indexing='ij')

    # Theta-ähnliche Struktur, symmetrisch
    A = (-1)**(ii + jj) / (ii + jj)**(beta)
    return A

# --- Parameter ---
N = 100  # Matrixgröße
beta_vals = [0.45, 0.48, 0.49, 0.5, 0.51, 0.52, 0.55]

# --- Analyse & Visualisierung ---
plt.figure(figsize=(10, 6))
for beta in beta_vals:
    H = construct_operator_matrix(N, beta)
    H_cpu = H.cpu().numpy()
    evals = np.linalg.eigvalsh(H_cpu)
    plt.plot(evals, '.', label=f'β = {beta:.2f}')

plt.axhline(0, color='gray', linestyle='--', alpha=0.7)
plt.title("Eigenwertspektren von $H_\\beta$ um die kritische Linie β = ½")
plt.xlabel("Index")
plt.ylabel("Eigenwert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
N = 100
betas = [0.45, 0.48, 0.49, 0.5, 0.51, 0.52, 0.55]

# Beispielhafte Eigenwert-Simulation (vereinfachtes Spektrum mit Paarungseigenschaft)
def eigenvalues_H_beta(beta, N):
    drift = (beta - 0.5) * 2.5  # einfache lineare Drift
    lambdas = np.linspace(-1, 1, N) + drift
    return np.sort(lambdas)

# Eigenwerte berechnen
eigen_dict = {beta: eigenvalues_H_beta(beta, N) for beta in betas}
ref = eigen_dict[0.5]

# Plot: Differenz zur kritischen Linie
fig, ax = plt.subplots(figsize=(10, 6))
for beta in betas:
    if beta == 0.5:
        continue
    delta = eigen_dict[beta] - ref
    ax.plot(range(N), delta, label=f'β = {beta:.2f} - Δ')

ax.axhline(0, color='gray', linestyle='--')
ax.set_title(r'Differenz der Eigenwerte $\lambda_\beta - \lambda_{0.5}$ zur kritischen Linie β = \frac{1}{2}')
ax.set_xlabel("Index")
ax.set_ylabel("Δ Eigenwert")
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# --- Setup ---
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

# --- Gerät auswählen ---
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Verwendetes Gerät: {device}")

# --- Daten einlesen ---
prime_data_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
zeta_data_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'

# Primzahlen einlesen
primes_raw = pd.read_csv(prime_data_path, header=None, dtype=str)[0]
primes_clean = pd.to_numeric(primes_raw, errors='coerce').dropna().astype(int).values[:2000]
log_primes = np.log(primes_clean)

# Zeta-Nullstellen einlesen
zeta_raw = pd.read_csv(zeta_data_path, header=None)[0].values[:2000]
zeta_norm = zeta_raw - np.mean(zeta_raw)

# --- Daten auf GPU übertragen ---
log_primes_tensor = torch.tensor(log_primes, dtype=torch.float32, device=device)
zeta_tensor = torch.tensor(zeta_norm, dtype=torch.float32, device=device)

# --- Fourier-Transformation ---
prime_spectrum = torch.fft.fft(log_primes_tensor)
zeta_spectrum = torch.fft.fft(zeta_tensor)

# --- Frequenzachse ---
N = len(zeta_tensor)
freqs = torch.fft.fftfreq(N, d=1.0)

# --- Ergebnisse auf CPU übertragen ---
prime_spectrum_cpu = prime_spectrum.abs().cpu().numpy()
zeta_spectrum_cpu = zeta_spectrum.abs().cpu().numpy()
freqs_cpu = freqs.cpu().numpy()

# --- Visualisierung ---
plt.figure(figsize=(12, 8))

# 1. Primzahlen
plt.subplot(3, 1, 1)
plt.plot(freqs_cpu[:N//2], prime_spectrum_cpu[:N//2], color='blue')
plt.title('Fourier-Spektrum der logarithmierten Primzahlen')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')

# 2. Zeta-Nullstellen
plt.subplot(3, 1, 2)
plt.plot(freqs_cpu[:N//2], zeta_spectrum_cpu[:N//2], color='green')
plt.title('Fourier-Spektrum der Zeta-Nullstellen')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')

# 3. Überlagerung
plt.subplot(3, 1, 3)
plt.plot(freqs_cpu[:N//2], prime_spectrum_cpu[:N//2], label='log(Primzahlen)', color='blue', alpha=0.6)
plt.plot(freqs_cpu[:N//2], zeta_spectrum_cpu[:N//2], label='Zeta-Nullstellen', color='green', alpha=0.6)
plt.title('Überlagerung der Spektren')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()

# --- Ähnlichkeitsmatrix ---
spectra_matrix = np.vstack([prime_spectrum_cpu[:N//2], zeta_spectrum_cpu[:N//2]])
similarity_matrix = cosine_similarity(spectra_matrix)

# Visualisierung der Ähnlichkeitsmatrix
plt.figure(figsize=(6, 5))
plt.imshow(similarity_matrix, cmap='viridis', interpolation='nearest')
plt.title('Spektrale Ähnlichkeitsmatrix\n(log(Primzahlen) vs. Zeta-Nullstellen)')
plt.colorbar(label='Kosinus-Ähnlichkeit')
plt.xticks([0, 1], ['log(Primzahlen)', 'Zeta-Nullstellen'])
plt.yticks([0, 1], ['log(Primzahlen)', 'Zeta-Nullstellen'])
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# GPU-Gerät wählen
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Gerät:", device)

# ----- Daten laden -----
prime_data_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
primes_raw = pd.read_csv(prime_data_path, header=None, dtype=str)[0]
primes_clean = pd.to_numeric(primes_raw, errors='coerce').dropna().astype(int).values[:100]
log_primes = torch.tensor(np.log(primes_clean), dtype=torch.float32, device=device)
log_primes_centered = log_primes - torch.mean(log_primes)

# ----- Simulierte Eigenwerte für β rund um 1/2 -----
betas = [0.45, 0.48, 0.49, 0.50, 0.51, 0.52, 0.55]
base_beta = 0.5
N = len(log_primes)

def simulate_eigenvalues(beta, size=N):
    torch.manual_seed(int(beta * 10000))
    base = torch.linspace(0, 2 * np.pi, size, device=device)
    shift = (beta - base_beta) * 50
    return torch.sin(base + shift) + torch.cos(base * beta)

eigenvalue_deltas = []
for beta in betas:
    ev = simulate_eigenvalues(beta)
    delta = ev - simulate_eigenvalues(base_beta)
    eigenvalue_deltas.append(delta.cpu().numpy())

# ----- Visualisierung -----
plt.figure(figsize=(14, 6))

# Plot der Eigenwertdifferenzen
for i, beta in enumerate(betas):
    label = rf"$\beta = {beta:.2f}$ - Δ"
    plt.plot(eigenvalue_deltas[i], label=label)

# Log-Primzahlen zentriert als Referenz
plt.scatter(range(N), log_primes_centered.cpu().numpy(), color='black', s=12, label='log(Primzahlen)', alpha=0.8)

plt.axhline(0, linestyle='--', color='gray')
plt.title("Vergleich von Operator-Eigenwerten und log(Primzahlen)")
plt.xlabel("Index")
plt.ylabel("Δ Eigenwert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---- GPU aktivieren ----
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Verwendetes Gerät:", device)

# ---- Datenpfade ----
prime_data_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
zeta_data_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'

# ---- Daten einlesen ----
primes_raw = pd.read_csv(prime_data_path, header=None, dtype=str)[0]
primes_clean = pd.to_numeric(primes_raw, errors='coerce').dropna().astype(int).values[:100]
log_primes = torch.tensor(np.log(primes_clean), dtype=torch.float32, device=device)
log_primes_centered = log_primes - torch.mean(log_primes)

zeta_raw = pd.read_csv(zeta_data_path, header=None)[0].values[:100]
zeta_centered = torch.tensor(zeta_raw, dtype=torch.float32, device=device)
zeta_centered -= torch.mean(zeta_centered)

# ---- Simulierte Operator-Eigenwerte ----
betas = [0.45, 0.48, 0.49, 0.50, 0.51, 0.52, 0.55]
base_beta = 0.5
N = len(log_primes)

def simulate_eigenvalues(beta, size=N):
    torch.manual_seed(int(beta * 10000))
    base = torch.linspace(0, 2 * np.pi, size, device=device)
    shift = (beta - base_beta) * 50
    return torch.sin(base + shift) + torch.cos(base * beta)

base_ev = simulate_eigenvalues(base_beta)
eigenvalue_deltas = [(simulate_eigenvalues(b) - base_ev).cpu().numpy() for b in betas]

# ---- Plot: Alles gemeinsam ----
plt.figure(figsize=(14, 6))

# Delta-Kurven
for i, beta in enumerate(betas):
    plt.plot(eigenvalue_deltas[i], label=rf"$\beta = {beta:.2f}$ - Δ")

# Log(Primzahlen) als Punkte
plt.scatter(range(N), log_primes_centered.cpu().numpy(), color='black', label='log(Primzahlen)', s=15, alpha=0.8, marker='x')

# Zeta-Nullstellen als Punkte
plt.scatter(range(N), zeta_centered.cpu().numpy(), color='purple', label='Zeta-Nullstellen', s=15, alpha=0.8, marker='o')

plt.axhline(0, linestyle='--', color='gray')
plt.title("Operator-Eigenwerte Δ vs. log(Primzahlen) & Zeta-Nullstellen")
plt.xlabel("Index")
plt.ylabel("Δ / Zentrierte Werte")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
