```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm
from statsmodels.tsa.stattools import acf

# Simulierte experimentelle Daten (ersetzen mit echten Daten)
N = np.arange(1000)  # Indexreihe
beta_exp = 0.5 + 0.02 * np.cos(2 * np.pi * 0.005 * N) + np.random.normal(0, 0.001, size=N.shape)

# Oszillierendes Modell definieren
def oscillating_model(N, A, B, C, D, E, F, G):
    return A + B * np.cos(C * N + D) + E * np.sin(F * N + G)

# Startwerte für die Optimierung
initial_params_osc = [0.5, 0.02, 0.03141593, 0, 0, 0, 0]

# Fit des Modells
params_osc, cov_osc = curve_fit(oscillating_model, N, beta_exp, p0=initial_params_osc)

# Berechnung des Fits
beta_fit_osc = oscillating_model(N, *params_osc)

# **Residualanalyse**
residuals = beta_exp - beta_fit_osc

# **Statistische Bewertung des Modells**
MSE_osc = np.mean(residuals**2)
AIC_osc = len(N) * np.log(MSE_osc) + 2 * len(params_osc)
BIC_osc = len(N) * np.log(MSE_osc) + len(params_osc) * np.log(len(N))

# **Visualisierungen**
plt.figure(figsize=(6,4))
plt.plot(N, beta_exp, label="Experimentelle β-Werte", color="blue")
plt.plot(N, beta_fit_osc, label="Oszillierendes Modell", linestyle="dashed", color="red")
plt.xlabel("Index N")
plt.ylabel("β(N)")
plt.title("Oszillierendes Modell vs. Experimentelle Daten")
plt.legend()
plt.show()

# **Histogramm der Residuen**
plt.figure(figsize=(6,4))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color="blue", label="Residuen")

# Fit einer Normalverteilung
mu, sigma = norm.fit(residuals)
x = np.linspace(min(residuals), max(residuals), 100)
plt.plot(x, norm.pdf(x, mu, sigma), color="red", label="Normalverteilung")

plt.xlabel("Residualwert")
plt.ylabel("Dichte")
plt.title("Histogramm der Residuen mit Normalverteilung")
plt.legend()
plt.show()

# **Autokorrelation der Residuen**
acf_values = acf(residuals, nlags=50, fft=True)
plt.figure(figsize=(6,4))
plt.stem(range(len(acf_values)), acf_values)
plt.fill_between(range(len(acf_values)), -1.96/np.sqrt(len(N)), 1.96/np.sqrt(len(N)), color="blue", alpha=0.2)
plt.axhline(0, linestyle="-", color="black")
plt.xlabel("Lag")
plt.ylabel("Autokorrelation")
plt.title("Autokorrelation der Residuen")
plt.show()

# **Fourier-Analyse der Residuen**
residuals_fft = np.fft.fft(residuals)
freqs = np.fft.fftfreq(len(residuals))

plt.figure(figsize=(6,4))
plt.plot(freqs[:len(freqs)//2], np.abs(residuals_fft[:len(freqs)//2]), color="green")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Residuen")
plt.show()

# **Ausgabe der optimierten Parameter und Modellbewertung**
print("\nOptimierte Parameter (Oszillierendes Modell):", params_osc)
print("MSE (Oszillierendes Modell):", MSE_osc)
print("AIC (Oszillierendes Modell):", AIC_osc)
print("BIC (Oszillierendes Modell):", BIC_osc)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import primepi
from scipy.optimize import curve_fit

# Feinstrukturkonstante
alpha = 1 / 137

# Freese-Formel: L(N) = A * N^(1 - beta) + E
# Wir testen mit einem angenäherten Wert für beta

def freese_formula(N, A, beta):
    return A * N**(1 - beta)

# Berechnung der Primzahldichte pi(N)
N_vals = np.linspace(10, 10**6, 500)
pi_vals = np.array([primepi(int(N)) for N in N_vals])

# Fit der Freese-Formel an pi(N)
popt, _ = curve_fit(freese_formula, N_vals, pi_vals)
A_opt, beta_opt = popt

# Berechnung der Skalierung
scaling_ratio = freese_formula(N_vals, A_opt, beta_opt) / pi_vals

# Überprüfung der Konvergenz gegen alpha
c_values = scaling_ratio / alpha

# Visualisierung der Ergebnisse
plt.figure(figsize=(10, 5))
plt.plot(N_vals, scaling_ratio, label='L(N)/π(N)', color='b')
plt.axhline(y=alpha, color='r', linestyle='--', label='Feinstrukturkonstante α')
plt.xlabel('N')
plt.ylabel('Skalierungsverhältnis')
plt.title('Vergleich von L(N) / π(N) mit der Feinstrukturkonstanten')
plt.legend()
plt.show()

print(f'Optimal gefittete Werte: A = {A_opt:.5f}, beta = {beta_opt:.5f}')
print(f'Mittleres Verhältnis c: {np.mean(c_values):.5f}')
```

```python
# 📌 Notwendige Bibliotheken importieren
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import pandas as pd

# 📌 1️⃣ Echte Nullstellen-Abstände aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"  # Falls der Dateiname abweicht, hier anpassen
try:
    zeros_data = np.loadtxt(file_path)  # Lade echte Nullstellen-Abstände
except FileNotFoundError:
    print(f"❌ Fehler: Datei '{file_path}' nicht gefunden. Stelle sicher, dass sie im richtigen Google Drive-Ordner liegt.")
    raise

# 📌 2️⃣ Erzeuge Index N passend zur Anzahl der Nullstellen
N = np.arange(1, len(zeros_data) + 1)
true_LN = zeros_data  # Verwende die echten Werte

# 📌 3️⃣ Stabilisierte Freese-Formel mit feinen Korrekturen
def freese_formula(N, A, beta, k, c1, c2, c3):
    return np.abs(A * N**beta * np.exp(k / N) * (1 + c1 * np.log(N) + c2 / N + c3 * np.sin(N)))

# 📌 4️⃣ Fehlerfunktion mit Sicherheitsfilter
def log_error_func(params, N, true_LN):
    A, beta, k, c1, c2, c3 = params
    predicted = freese_formula(N, A, beta, k, c1, c2, c3)

    # Sicherstellen, dass alle Werte positiv sind
    valid_indices = (predicted > 0) & (true_LN > 0)
    if np.sum(valid_indices) == 0:
        return np.inf  # Falls alle Werte ungültig sind, extrem hoher Fehler

    return np.mean((np.log(predicted[valid_indices]) - np.log(true_LN[valid_indices]))**2)

# 📌 5️⃣ Begrenzung der Datenmenge für bessere Konvergenz
sample_size = min(50000, len(N))  # Maximal 50.000 Werte verwenden
N_reduced = N[:sample_size]
true_LN_reduced = true_LN[:sample_size]

# 📌 6️⃣ Optimierungsgrenzen & Startwerte
initial_params = [1.5, 0.1, 1.8, 0.01, -30, 0.001]
bounds = [
    (0.5, 5),       # A (Skalierung erweitern)
    (-0.5, 0.5),    # beta (erhöhte Wachstumsrate zulassen)
    (0.1, 5),       # k (Exponentialterm erweitern)
    (-0.1, 0.1),    # c1 (Logarithmische Korrektur feinjustieren)
    (-50, 50),      # c2 (Erweiterte Korrektur)
    (-0.01, 0.01)   # c3 (Sinus-Korrektur feinjustieren)
]

# 📌 7️⃣ Optimierung mit Differential Evolution
print("🚀 Starte Optimierung mit Differential Evolution... (dauert einige Minuten)")
opt_params = opt.differential_evolution(log_error_func, bounds=bounds, args=(N_reduced, true_LN_reduced)).x
print("✅ Optimierung abgeschlossen!")

# 📌 8️⃣ Extrahiere die optimalen Parameter
A_opt, beta_opt, k_opt, c1_opt, c2_opt, c3_opt = opt_params

# 📌 9️⃣ Berechnete Werte für Visualisierung
fitted_LN_reduced = freese_formula(N_reduced, A_opt, beta_opt, k_opt, c1_opt, c2_opt, c3_opt)

# 📌 🔟 Ergebnisse speichern
optimal_results = {
    "A": A_opt,
    "beta": beta_opt,
    "k": k_opt,
    "c1": c1_opt,
    "c2": c2_opt,
    "c3": c3_opt
}

# 📌 1️⃣1️⃣ Visualisierung der Nullstellen-Abstände mit stabilisierter Formel
plt.figure(figsize=(10, 6))
plt.scatter(N_reduced, true_LN_reduced, s=1, label="Echte Nullstellen-Abstände", alpha=0.5)
plt.plot(N_reduced, fitted_LN_reduced, 'r--', linewidth=1.5, label="Optimierte Freese-Formel (stabilisiert)")
plt.xlabel("Index N")
plt.ylabel("L(N) - Nullstellen-Abstände")
plt.title("Freese-Skalierung vs. Nullstellen (stabilisiert)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# 📌 1️⃣2️⃣ Optimale Parameter ausgeben
df_results = pd.DataFrame(list(optimal_results.items()), columns=["Parameter", "Wert"])
print("\n✅ Optimierte Parameter der stabilisierten Freese-Formel:")
display(df_results)
```

```python
import numpy as np
from scipy.sparse import diags
from scipy.sparse.linalg import eigs

# Parameter
N = 1000      # Anzahl Gitterpunkte
L = 50.0      # Ortsbereich: x ∈ [-L, L]
lambda_val = 0.1  # Stärke des x²-Potentials

# Ortsgitter und Impulsoperator (FFT-basiert)
x = np.linspace(-L, L, N, endpoint=False)  # Diskretes x-Gitter
dx = x[1] - x[0]  # Gitterabstand

# Konstruiere den Impulsoperator p = -iℏ ∂/∂x (periodische Randbedingungen)
p_fft = np.fft.fftfreq(N, d=dx) * 2 * np.pi  # ℏ = 1

# Diagonalmatrizen für x und p
X = diags(x)                     # Ortsoperator
P = diags(p_fft)                 # Impulsoperator (Fourier-Raum)
H_kin = 0.5 * (X @ P + P @ X)    # Symmetrisierter kinetischer Term (xp + px)
H_pot = lambda_val * X @ X       # Potentialterm λx²

# Gesamt-Hamiltonian (als sparse Matrix)
H = H_kin + H_pot

# Berechne die 100 kleinsten Eigenwerte
eigenvalues, _ = eigs(H, k=100, sigma=0)  # sigma=0 → Eigenwerte nahe 0
sorted_eigs = np.sort(np.real(eigenvalues))

# Ausgabe der ersten 10 Eigenwerte
print("Kleinste Eigenwerte des modifizierten Berry-Keating-Hamiltonians:")
print(sorted_eigs[:10])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
from scipy.fftpack import fft
from statsmodels.tsa.stattools import acf

# Experimentelle Werte für Beta (simuliert)
N = np.arange(1, 1001)
beta_exp = 0.5 + 0.02 * np.cos(2 * np.pi * N / 100) + 0.001 * np.random.randn(len(N))

# Theoretische Beziehung für Beta: β = 1 - φ/π
phi = (1 + np.sqrt(5)) / 2  # Goldener Schnitt
theoretical_beta = 1 - phi / np.pi

print(f"Theoretischer Beta-Wert: {theoretical_beta}")

# Curve Fitting: Modell für Beta-Schwingungen
def oscillating_model(n, A, B, C, D):
    return A + B * np.cos(C * n + D)

initial_params = [0.5, 0.02, 2*np.pi/100, 0]
params_opt, _ = curve_fit(oscillating_model, N, beta_exp, p0=initial_params)

print(f"Optimierte Parameter: {params_opt}")

# Plot: Experimentelle vs. Modellierte Beta-Werte
plt.figure(figsize=(7,5))
plt.plot(N, beta_exp, 'b-', label='Experimentelle β-Werte')
plt.plot(N, oscillating_model(N, *params_opt), 'r--', label='Fitted β-Modell')
plt.xlabel("Index N")
plt.ylabel("β(N)")
plt.title("Oszillierendes Modell für Beta-Werte")
plt.legend()
plt.show()

# Fourier-Analyse der Beta-Werte
beta_fft = np.abs(fft(beta_exp))[:len(N)//2]
frequencies = np.linspace(0, 0.5, len(beta_fft))

plt.figure(figsize=(7,5))
plt.plot(frequencies, beta_fft, 'g-', label='Fourier-Spektrum')
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Stärke")
plt.title("Fourier-Analyse der Beta-Werte")
plt.legend()
plt.show()

# Autokorrelationsanalyse
acf_values = acf(beta_exp, nlags=50, fft=True)
plt.figure(figsize=(6,4))
plt.stem(range(len(acf_values)), acf_values)
plt.axhline(0, linestyle='-', color='black')
plt.title("Autokorrelation der Beta-Werte")
plt.xlabel("Lag")
plt.ylabel("Autokorrelation")
plt.show()

# Histogramm zur Normalverteilung der Residuen
residuals = beta_exp - oscillating_model(N, *params_opt)
plt.figure(figsize=(7,5))
plt.hist(residuals, bins=30, color='b', alpha=0.7, label='Residuen')
plt.xlabel("Residualwert")
plt.ylabel("Dichte")
plt.title("Histogramm der Residuen mit Normalverteilung")
plt.legend()
plt.show()
```
