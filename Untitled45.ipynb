```python
# 📌 Wichtige Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq

# 📌 1️⃣ Google Drive einbinden und Datei laden
from google.colab import drive
drive.mount('/content/drive')

# 📌 2️⃣ Datei laden
file_path = "/content/drive/MyDrive/beta_scale_operator.csv"
df = pd.read_csv(file_path)

# 📌 3️⃣ Überprüfen der Spaltennamen
print("📊 Verfügbare Spalten:", df.columns)
if 'N' not in df.columns or 'beta_scale' not in df.columns:
    raise ValueError("⚠ Erwartete Spalten ('N', 'beta_scale') nicht gefunden!")

# 📌 4️⃣ Daten für Modellierung extrahieren
N_values = df['N'].values  # Index der Nullstellen
beta_values = df['beta_scale'].values  # Entsprechende Beta-Werte

# 📌 5️⃣ Universelle Modellfunktion für β(N)
def beta_model(N, B0, B1, B2, omega):
    return B0 + B1 / np.log(N) + B2 * np.cos(omega * np.log(N))

# 📌 6️⃣ Curve Fit durchführen
popt, pcov = curve_fit(beta_model, N_values, beta_values, maxfev=10000)
B0_opt, B1_opt, B2_opt, omega_opt = popt
print(f"✅ Optimierte Parameter für β(N): B0={B0_opt:.6f}, B1={B1_opt:.6f}, B2={B2_opt:.6f}, ω={omega_opt:.6f}")

# 📌 7️⃣ Visualisierung des Fits
plt.figure(figsize=(8,5))
plt.scatter(N_values, beta_values, s=1, label="Beta-Skala (Daten)", alpha=0.5)
plt.plot(N_values, beta_model(N_values, *popt), 'r-', label="Fit: β(N)", linewidth=2)
plt.xlabel("N (Zeta-Nullstellen-Index)")
plt.ylabel("β(N)")
plt.xscale("log")
plt.title("Anpassung von β(N)")
plt.legend()
plt.grid()
plt.show()

# 📌 8️⃣ Erweiterte Freese-Formel mit β(N)
def freese_formula(N, A, C1, C2):
    beta_N = beta_model(N, *popt)
    return A * N**beta_N + C1 * np.log(N) + C2

# 📌 9️⃣ Fehlerbewertung
L_Freese = freese_formula(N_values, A=2.142762, C1=-0.292849, C2=6.270631)
mse = np.mean((L_Freese - beta_values)**2)
print(f"📊 Fehlerbewertung (MSE) der erweiterten Formel: {mse:.6f}")

# 📌 🔟 (Optional) Fourier-Analyse von β(N) zur Entdeckung von Frequenzmustern
beta_fft = fft(beta_values)
freqs = fftfreq(len(N_values))

plt.figure(figsize=(8,5))
plt.plot(freqs[:len(freqs)//2], np.abs(beta_fft[:len(freqs)//2]), 'b-')
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektrum von β(N)")
plt.grid()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib torch requests

# 📌 Bibliotheken importieren
import numpy as np
import torch
import requests
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from io import StringIO

# 📌 Google Drive Links zu den Datensätzen
zeta_url = "https://drive.google.com/uc?export=download&id=1KInLS9p4FHwV-WgDwhZ0ahOCqzk04-61"
beta_url = "https://drive.google.com/uc?export=download&id=1Zkz8ycT_HnHzP_rUsPtLlU93mu1qyICA"
freese_url = "https://drive.google.com/uc?export=download&id=1-Oho0uYq7PPhTyyGq8e_B_FQLeMP61yy"

# 📌 Funktion zum Laden der Daten
def load_data(url):
    response = requests.get(url)
    data = np.loadtxt(StringIO(response.text), delimiter=",", skiprows=1)  # Header ignorieren
    return torch.tensor(data, dtype=torch.float32)

# 📥 Lade alle Datensätze herunter
print("📥 Lade Zeta-Nullstellen, Beta-Werte & Freese-Funktion herunter...")
zeta_zeros = load_data(zeta_url)[:10_000]   # Speicheroptimierung: Nur 10.000 Werte
beta_values = load_data(beta_url)[:10_000]
freese_values = load_data(freese_url)[:10_000]

print(f"✅ Erfolgreich {len(zeta_zeros)} Zeta-Nullstellen, {len(beta_values)} Beta-Werte & {len(freese_values)} Freese-Werte geladen!")

# 📌 Berechnung der Ableitungen (numerische Differentiation)
def numerical_derivative(values):
    return torch.diff(values) / torch.diff(torch.arange(len(values), dtype=torch.float32))

zeta_derivative = numerical_derivative(zeta_zeros)
beta_derivative = numerical_derivative(beta_values)
freese_derivative = numerical_derivative(freese_values)

# 📌 Fourier-Analyse
def fourier_analysis(data):
    spectrum = torch.fft.fft(data).abs()
    frequencies = torch.fft.fftfreq(len(data))
    return frequencies, spectrum

freq_zeta, fft_zeta = fourier_analysis(zeta_zeros)
freq_beta, fft_beta = fourier_analysis(beta_values)
freq_freese, fft_freese = fourier_analysis(freese_values)

# 📌 Plot der Funktionen
plt.figure(figsize=(10,5))
plt.plot(zeta_zeros[:500], label="Zeta-Nullstellen", color="blue")
plt.plot(beta_values[:500], label="Beta-Skala", color="red")
plt.plot(freese_values[:500], label="Freese-Funktion", color="green")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Vergleich: Zeta-Nullstellen, Beta-Werte & Freese-Funktion")
plt.grid()
plt.show()

# 📌 Plot der Ableitungen
plt.figure(figsize=(10,5))
plt.plot(zeta_derivative[:500], label="Zeta-Ableitung", color="blue")
plt.plot(beta_derivative[:500], label="Beta-Ableitung", color="red")
plt.plot(freese_derivative[:500], label="Freese-Ableitung", color="green")
plt.xlabel("Index")
plt.ylabel("Ableitung")
plt.legend()
plt.title("Numerische Ableitungen")
plt.grid()
plt.show()

# 📌 Plot der Fourier-Analysen
plt.figure(figsize=(10,5))
plt.plot(freq_zeta[:len(freq_zeta)//2], fft_zeta[:len(fft_zeta)//2], "b-", label="Zeta-FFT")
plt.plot(freq_beta[:len(freq_beta)//2], fft_beta[:len(fft_beta)//2], "r-", label="Beta-FFT")
plt.plot(freq_freese[:len(freq_freese)//2], fft_freese[:len(fft_freese)//2], "g-", label="Freese-FFT")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektren")
plt.legend()
plt.grid()
plt.show()
```

```python
# 🚀 Fourier-Analyse der Freese-Reihe + Vergleich mit Zeta-Nullstellen
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq

# 📌 Parameter
N = 1000  # Anzahl der Punkte
x = np.linspace(0.1, 10, N)

# 📌 Definition der Freese-Reihe
def freese_reihe(x, beta):
    return x**(-beta)

# 📌 Beta-Werte aus vorheriger Analyse
beta_1 = 0.484906
beta_2 = 0.492206

# 📌 Berechnung der Werte für die Freese-Reihe
F1 = freese_reihe(x, beta_1)
F2 = freese_reihe(x, beta_2)

# 🚀 Fourier-Transformation
freqs = fftfreq(N, d=(x[1] - x[0]))  # Frequenzachsen berechnen
fft_F1 = np.abs(fft(F1))  # Fourier-Spektrum für β1
fft_F2 = np.abs(fft(F2))  # Fourier-Spektrum für β2

# 📊 Visualisierung der Freese-Reihe
plt.figure(figsize=(8, 5))
plt.plot(x, F1, label=f"$\\beta_1 = {beta_1}$", color="blue")
plt.plot(x, F2, "--", label=f"$\\beta_2 = {beta_2}$", color="red")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Freese-Reihe mit verschiedenen Beta-Werten")
plt.legend()
plt.show()

# 📊 Visualisierung der Fourier-Spektren
plt.figure(figsize=(8, 5))
plt.plot(freqs[:N//2], fft_F1[:N//2], label=f"Fourier $\\beta_1 = {beta_1}$", color="blue")
plt.plot(freqs[:N//2], fft_F2[:N//2], "--", label=f"Fourier $\\beta_2 = {beta_2}$", color="red")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Analyse der Freese-Reihe")
plt.legend()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
from scipy.fft import fft, fftfreq

# 📌 URL der Primzahlen-Datei
file_url = "https://drive.google.com/uc?id=1f8U4hxpgbgGvfIiyNWSCrik8NZI5RBUK"

# 📌 Datei herunterladen und speichern
filename = "primzahlen.txt"
response = requests.get(file_url)
with open(filename, "wb") as f:
    f.write(response.content)

# 📌 Datei einlesen (Erwartet eine Spalte von Primzahlen)
primes = np.loadtxt(filename, dtype=int)

# 📌 Modulo-Verteilungen für verschiedene Basiswerte
moduli = [3, 7, 21, 33, 137, 33300]
mod_distributions = {m: np.mod(primes, m) for m in moduli}

# 📌 Fourier-Analyse der Verteilungen
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.ravel()

for i, m in enumerate(moduli):
    freqs = fftfreq(len(primes))
    spectrum = np.abs(fft(mod_distributions[m]))

    axes[i].plot(freqs[:len(freqs)//2], spectrum[:len(freqs)//2], label=f"Modulo {m}")
    axes[i].set_title(f"Frequenzspektrum (Modulo {m})")
    axes[i].set_xlabel("Frequenz")
    axes[i].set_ylabel("Amplitude")
    axes[i].legend()

plt.tight_layout()
plt.show()

# 📌 Vergleich mit Beta-Skala-Frequenzen
beta_freqs = [1/33, 1/137, 7/33300]
plt.figure(figsize=(8,5))
plt.plot(freqs[:len(freqs)//2], spectrum[:len(freqs)//2], label="Primzahlen Fourier-Transform", color="black")

# 📌 Beta-Frequenzen als Referenzlinien
for bf in beta_freqs:
    plt.axvline(bf, linestyle="dashed", label=f"Beta-Frequenz {bf}")

plt.title("Frequenzvergleich: Primzahlen vs. Beta-Skala")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```

```python
# ✅ Notwendige Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error

# ✅ Google Drive einbinden (falls notwendig)
from google.colab import drive
drive.mount('/content/drive')

# ✅ Datenpfade
beta_file = "/content/drive/MyDrive/beta_scale_operator.csv"
error_file = "/content/drive/MyDrive/errors_oscillation.csv"

# ✅ Daten laden
df_beta = pd.read_csv(beta_file)
df_error = pd.read_csv(error_file)

# ✅ Zusammenführen nach 'Nullstellen'
df = pd.merge(df_beta, df_error, on="Nullstellen")

# 📌 1️⃣ Lineares Modell: Error_Oscillation = a * Beta_Scale + b
def linear_model(x, a, b):
    return a * x + b

popt_linear, _ = curve_fit(linear_model, df["Beta_Scale"], df["Error_Oscillation"])
mse_linear = mean_squared_error(df["Error_Oscillation"], linear_model(df["Beta_Scale"], *popt_linear))
r2_linear = linregress(df["Beta_Scale"], df["Error_Oscillation"]).rvalue**2

# 📌 2️⃣ Logarithmisches Modell: Error_Oscillation = a * log(Beta_Scale) + b
def log_model(x, a, b):
    return a * np.log(x) + b

popt_log, _ = curve_fit(log_model, df["Beta_Scale"], df["Error_Oscillation"])
mse_log = mean_squared_error(df["Error_Oscillation"], log_model(df["Beta_Scale"], *popt_log))

# 📌 3️⃣ Potenzgesetz: Error_Oscillation = A * Beta_Scale^B + C
def power_model(x, A, B, C):
    return A * x**B + C

popt_power, _ = curve_fit(power_model, df["Beta_Scale"], df["Error_Oscillation"])
mse_power = mean_squared_error(df["Error_Oscillation"], power_model(df["Beta_Scale"], *popt_power))

# 📌 4️⃣ Fourier-Korrekturmodell: Error_Oscillation = A * Beta_Scale^B + C + D * cos(ω log(N))
def fourier_model(x, A, B, C, D, ω):
    return A * x**B + C + D * np.cos(ω * np.log(x))

popt_fourier, _ = curve_fit(fourier_model, df["Beta_Scale"], df["Error_Oscillation"])
mse_fourier = mean_squared_error(df["Error_Oscillation"], fourier_model(df["Beta_Scale"], *popt_fourier))

# ✅ Ergebnisse ausgeben
print(f"📊 Lineares Modell: Error = {popt_linear[0]:.6f} * Beta + {popt_linear[1]:.6f}  (MSE = {mse_linear:.6f}, R² = {r2_linear:.6f})")
print(f"📊 Logarithmisches Modell: Error = {popt_log[0]:.6f} * log(Beta) + {popt_log[1]:.6f}  (MSE = {mse_log:.6f})")
print(f"📊 Potenzgesetz: Error = {popt_power[0]:.6f} * Beta^{popt_power[1]:.6f} + {popt_power[2]:.6f}  (MSE = {mse_power:.6f})")
print(f"📊 Fourier-Korrekturmodell: Error = {popt_fourier[0]:.6f} * Beta^{popt_fourier[1]:.6f} + {popt_fourier[2]:.6f} + {popt_fourier[3]:.6f} * cos({popt_fourier[4]:.6f} log(N))  (MSE = {mse_fourier:.6f})")

# ✅ Beste Funktion anhand der geringsten MSE bestimmen
best_model = min([(mse_linear, "Linear"), (mse_log, "Log"), (mse_power, "Power"), (mse_fourier, "Fourier")], key=lambda x: x[0])
print(f"\n🏆 Beste Fit-Funktion: {best_model[1]} (MSE = {best_model[0]:.6f})")

# 📊 Visualisierung des besten Fits
plt.figure(figsize=(8, 6))
plt.scatter(df["Beta_Scale"], df["Error_Oscillation"], s=5, alpha=0.5, label="Datenpunkte")

x_range = np.linspace(min(df["Beta_Scale"]), max(df["Beta_Scale"]), 1000)

if best_model[1] == "Linear":
    plt.plot(x_range, linear_model(x_range, *popt_linear), "r-", label="Linearer Fit")
elif best_model[1] == "Log":
    plt.plot(x_range, log_model(x_range, *popt_log), "g-", label="Logarithmischer Fit")
elif best_model[1] == "Power":
    plt.plot(x_range, power_model(x_range, *popt_power), "b-", label="Potenzgesetz Fit")
elif best_model[1] == "Fourier":
    plt.plot(x_range, fourier_model(x_range, *popt_fourier), "m-", label="Fourier Fit")

plt.xlabel("Beta_Scale")
plt.ylabel("Error_Oscillation")
plt.title("Beste Fit-Funktion für Error_Oscillation")
plt.legend()
plt.grid()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
from scipy.optimize import curve_fit

# 📌 Datei aus Google Drive laden
file_path = "/content/drive/MyDrive/errors_oscillation.csv"

# 📌 Daten einlesen
df = pd.read_csv(file_path)
df.dropna(inplace=True)  # Sicherheitscheck gegen NaN-Werte
error_values = df["Error_Oscillation"].values
N = len(error_values)

# 📌 Fourier-Analyse
frequencies = fftfreq(N)
fft_values = fft(error_values)
dominant_indices = np.argsort(np.abs(fft_values))[-5:]  # Top 5 Frequenzen
dominant_frequencies = frequencies[dominant_indices]
dominant_amplitudes = np.abs(fft_values[dominant_indices])

print("✅ Dominante Frequenzen des Fehlers:", np.round(dominant_frequencies, 10))
print("✅ Korrespondierende Amplituden:", np.round(dominant_amplitudes, 2))

# 📌 Modell für Oszillation definieren (Kosinus-Ansatz)
def oscillation_model(n, A, f, phi, C):
    return A * np.cos(2 * np.pi * f * n + phi) + C

# 📌 Kosinus-Fit mit dominanter Frequenz durchführen
dominant_freq = dominant_frequencies[-1]  # Stärkste Frequenz wählen
initial_guess = [np.max(error_values), dominant_freq, 0, np.mean(error_values)]
params, _ = curve_fit(oscillation_model, np.arange(N), error_values, p0=initial_guess)

# 📌 Modellierte Oszillation berechnen
fitted_oscillation = oscillation_model(np.arange(N), *params)

# 📌 Plots erstellen
plt.figure(figsize=(10, 5))
plt.plot(df["Nullstellen"], error_values, label="Original-Fehler", alpha=0.6)
plt.plot(df["Nullstellen"], fitted_oscillation, 'r--', label="Fourier-Fit (Oszillation)")
plt.xlabel("Nullstellen-Index")
plt.ylabel("Fehler-Oszillation")
plt.title("Fourier-Modell der Oszillation im Fehler")
plt.legend()
plt.grid()
plt.show()

# 📌 Ergebnisse ausgeben
print(f"✅ Optimierte Oszillations-Parameter:")
print(f"   Amplitude: {params[0]:.6f}")
print(f"   Frequenz: {params[1]:.10f}")
print(f"   Phase: {params[2]:.6f}")
print(f"   Offset: {params[3]:.6f}")
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib sympy

# 📌 Importiere Bibliotheken
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
from sympy import symbols, solve

# 📌 Zeta-Nullstellen aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"

try:
    with open(file_path, "r") as f:
        zeta_zeros = np.array([float(line.strip()) for line in f.readlines() if line.strip()])[:1000]
    print("✅ Erfolgreich 1000 echte Zeta-Nullstellen geladen!")
except Exception as e:
    print(f"❌ Fehler beim Laden der Datei: {e}")
    zeta_zeros = np.linspace(0, 1000, 1000)  # Fallback: Dummy-Daten

# 📌 Hamilton-Operator definieren (Matrix)
N = len(zeta_zeros)
H = np.diag(zeta_zeros) + np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-1), k=-1)

# 📌 Eigenwerte berechnen
eigenvalues = np.linalg.eigvalsh(H)

# 📌 Analytischer Fit-Ansatz (z.B. logarithmische Skala)
def model(n, a, b):
    return a * np.log(n) + b

# 📌 Fit-Optimierung durchführen
n_values = np.arange(1, len(eigenvalues) + 1)
params, cov = opt.curve_fit(model, n_values, eigenvalues)

# 📌 Fit-Parameter ausgeben
a_opt, b_opt = params
print(f"📊 Optimierte Parameter: a={a_opt:.6f}, b={b_opt:.6f}")

# 📌 Fehlerbewertung (MSE, R²)
mse = np.mean((model(n_values, *params) - eigenvalues)**2)
r2 = np.corrcoef(model(n_values, *params), eigenvalues)[0, 1]**2
print(f"📊 Fehlerbewertung: MSE={mse:.6f}, R²={r2:.6f}")

# 📌 Plot: Vergleich Eigenwerte vs. Fit
plt.figure(figsize=(8,5))
plt.plot(n_values, eigenvalues, "bo", markersize=2, label="Eigenwerte (H)")
plt.plot(n_values, model(n_values, *params), "r-", label=f"Fit: a log(n) + b (R²={r2:.4f})")
plt.xlabel("Index n")
plt.ylabel("Eigenwerte")
plt.legend()
plt.title("Fit-Optimierung der Operator-Eigenwerte")
plt.grid()
plt.show()
```

```python
# 📌 Erweiterter Modellansatz: Logarithmus + Potenz + Oszillation
def extended_model(n, a, b, c, d, A, B):
    return a * np.log(n) + b + c * (n ** d) + A * np.cos(B * np.log(n))

# 📌 Fit-Optimierung durchführen
params_ext, cov_ext = opt.curve_fit(extended_model, n_values, eigenvalues,
                                    p0=[353, -1298, 1, 0.1, 10, 0.5])

# 📌 Optimierte Parameter ausgeben
a_opt, b_opt, c_opt, d_opt, A_opt, B_opt = params_ext
print(f"📊 Erweiterte Fit-Parameter: a={a_opt:.6f}, b={b_opt:.6f}, c={c_opt:.6f}, d={d_opt:.6f}, A={A_opt:.6f}, B={B_opt:.6f}")

# 📌 Fehlerbewertung (MSE, R²)
mse_ext = np.mean((extended_model(n_values, *params_ext) - eigenvalues)**2)
r2_ext = np.corrcoef(extended_model(n_values, *params_ext), eigenvalues)[0, 1]**2
print(f"📊 Fehlerbewertung (erweitertes Modell): MSE={mse_ext:.6f}, R²={r2_ext:.6f}")

# 📌 Plot: Vergleich Eigenwerte vs. Fit
plt.figure(figsize=(8,5))
plt.plot(n_values, eigenvalues, "bo", markersize=2, label="Eigenwerte (H)")
plt.plot(n_values, extended_model(n_values, *params_ext), "r-", label=f"Erweiterter Fit (R²={r2_ext:.4f})")
plt.xlabel("Index n")
plt.ylabel("Eigenwerte")
plt.legend()
plt.title("Erweiterter Fit mit Oszillationen")
plt.grid()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib

# 📌 Importiere Bibliotheken
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
import torch

# 📌 GPU-Check
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🚀 Berechnung auf: {device}")

# 📥 Zeta-Nullstellen von Google Drive laden (10.000 Stück)
zeta_url = "https://drive.google.com/uc?id=1KInLS9p4FHwV-WgDwhZ0ahOCqzk04-61"

try:
    zeta_zeros = np.loadtxt(zeta_url, max_rows=10000)  # Direkt aus URL
    print(f"✅ Erfolgreich {len(zeta_zeros)} Zeta-Nullstellen geladen!")
except Exception as e:
    print(f"❌ Fehler beim Laden der Datei: {e}")
    zeta_zeros = np.linspace(0, 10000, 10000)  # Fallback: Dummy-Daten

# 📌 Hamilton-Operator definieren (Sparse-Matrix für GPU)
N = len(zeta_zeros)
diagonal_terms = torch.tensor(zeta_zeros, dtype=torch.float32, device=device)

# Sparse Matrix mit Neben- und Hauptdiagonale
H = torch.sparse_coo_tensor(
    indices=torch.tensor([
        list(range(N)) + list(range(N-1)) + list(range(1, N)),
        list(range(N)) + list(range(1, N)) + list(range(N-1))
    ], dtype=torch.int64, device=device),
    values=torch.cat([
        diagonal_terms,                # Hauptdiagonale
        torch.ones(N-1, device=device),  # Oberdiagonale
        torch.ones(N-1, device=device)   # Unterdiagonale
    ]),
    size=(N, N)
)

# ✅ Selbstadjungiertheit überprüfen
H_dense = H.to_dense()
H_dagger = H_dense.T.conj()
is_self_adjoint = torch.allclose(H_dense, H_dagger, atol=1e-10)
print(f"✅ Selbstadjungiert: {is_self_adjoint}")

# 🔥 Berechne nur die **10 größten** Eigenwerte (Speicheroptimierung)
eigenvalues, _ = torch.lobpcg(H_dense, k=10)
print(f"🔥 10 Größte Eigenwerte: {eigenvalues.cpu().numpy()}")

# 📌 Fourier-Analyse der Eigenwerte
fft_spectrum = torch.fft.fft(eigenvalues)
frequencies = torch.fft.fftfreq(len(eigenvalues))

# 📌 Dominante Frequenzen extrahieren
dominant_indices = torch.argsort(torch.abs(fft_spectrum))[-5:]  # Top 5 Frequenzen
dominant_frequencies = frequencies[dominant_indices].cpu().numpy()
dominant_amplitudes = torch.abs(fft_spectrum[dominant_indices]).cpu().numpy()

print(f"🔥 Dominante Frequenzen: {dominant_frequencies}")
print(f"🔥 Korrespondierende Amplituden: {dominant_amplitudes}")

# ✅ Euler-Freese-Fehlerwert berechnen
euler_freese_error = torch.mean((eigenvalues - torch.tensor(zeta_zeros[-10:], device=device)) ** 2).cpu().numpy()
print(f"✅ Euler-Freese-Fehlerwert: {euler_freese_error:.5f}")

# 📊 Plot: Eigenwerte vs. Index
plt.figure(figsize=(7, 5))
plt.plot(eigenvalues.cpu().numpy(), "bo-", markersize=3, label="Eigenwerte")
plt.xlabel("Index")
plt.ylabel("Eigenwerte")
plt.title("Spektrum des Hamilton-Operators")
plt.legend()
plt.grid()
plt.show()

# 📊 Plot: Fourier-Spektrum der Eigenwerte
plt.figure(figsize=(7, 5))
plt.plot(frequencies.cpu().numpy(), np.abs(fft_spectrum.cpu().numpy()), "r-")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektrum der Eigenwerte")
plt.grid()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken installieren (falls noch nicht installiert)
!pip install numpy scipy torch matplotlib

# 📌 Importiere Bibliotheken
import torch
import numpy as np
import matplotlib.pyplot as plt

# 📌 Berechnungsgerät (GPU, falls verfügbar)
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"🚀 Berechnung auf: {device}")

# 📌 Pfade zu den Daten auf Google Drive
zeta_path = "/content/drive/MyDrive/zeros6.txt"
beta_path = "/content/drive/MyDrive/beta_scale_operator.csv"
freese_path = "/content/drive/MyDrive/freese_function_odlyzko_cleaned.csv"

# 📌 Maximale Anzahl der Datenpunkte
N = 10_000  # Speicheroptimierung: Nur 10.000 Werte verwenden


# 📌 Funktion zum Laden von Daten mit Fehlerkorrektur
def load_data(file_path, N, name):
    try:
        with open(file_path, "r") as f:
            raw_lines = f.readlines()

        numeric_lines = []
        for line in raw_lines:
            line = line.strip().replace(",", ".")  # Dezimaltrennung anpassen
            if line == "" or not any(c.isdigit() for c in line):
                continue  # Leerzeilen und nicht-numerische Zeilen ignorieren
            try:
                value = float(line)
                numeric_lines.append(value)
            except ValueError:
                print(f"⚠️ Warnung: Problematische Zeile in {name}: {line} → wird ignoriert.")

        if len(numeric_lines) == 0:
            print(f"❌ Fehler: Datei für {name} enthält keine nutzbaren Zahlen. Ersetze durch Dummy-Daten!")
            return torch.linspace(0.1, 10, N, device=device)  # Dummy-Daten als Fallback

        return torch.tensor(numeric_lines[:N], dtype=torch.float32, device=device)

    except Exception as e:
        print(f"❌ Fehler beim Laden der Datei {name}: {e}")
        return torch.linspace(0.1, 10, N, device=device)  # Dummy-Daten als Fallback


# 📥 Lade Zeta-Nullstellen, Beta-Werte & Freese-Werte
print("📥 Lade Zeta-Nullstellen, Beta-Werte & Freese-Werte herunter...")
zeta_zeros = load_data(zeta_path, N, "Zeta-Nullstellen")
beta_values = load_data(beta_path, N, "Beta-Skala")
freese_values = load_data(freese_path, N, "Freese-Werte")

print(f"✅ Erfolgreich {len(zeta_zeros)} Zeta-Nullstellen, {len(beta_values)} Beta-Werte & {len(freese_values)} Freese-Werte geladen!")


# 📌 Fourier-Analyse-Funktion
def fourier_analysis(data):
    fft_spectrum = torch.fft.fft(data)
    frequencies = torch.fft.fftfreq(len(data), d=1.0)
    return frequencies, fft_spectrum


# 📌 Berechne Fourier-Transformationen
freq_zeta, fft_zeta = fourier_analysis(zeta_zeros)
freq_beta, fft_beta = fourier_analysis(beta_values)
freq_freese, fft_freese = fourier_analysis(freese_values)


# 📌 Funktion zur Extraktion dominanter Frequenzen
def dominant_freqs(fft_spectrum, frequencies):
    indices = torch.argsort(torch.abs(fft_spectrum))[-5:]  # Top 5 Frequenzen
    return frequencies[indices].cpu().numpy(), torch.abs(fft_spectrum[indices]).cpu().numpy()


# 📌 Dominante Frequenzen extrahieren
freq_dom_zeta, amp_dom_zeta = dominant_freqs(fft_zeta, freq_zeta)
freq_dom_beta, amp_dom_beta = dominant_freqs(fft_beta, freq_beta)
freq_dom_freese, amp_dom_freese = dominant_freqs(fft_freese, freq_freese)


# 📌 Ergebnisse ausgeben
print("\n🔥 Dominante Frequenzen der Zeta-Nullstellen:", np.round(freq_dom_zeta, 6))
print("🔥 Dominante Amplituden der Zeta-Nullstellen:", np.round(amp_dom_zeta, 2))

print("\n🔥 Dominante Frequenzen der Beta-Skala:", np.round(freq_dom_beta, 6))
print("🔥 Dominante Amplituden der Beta-Skala:", np.round(amp_dom_beta, 2))

print("\n🔥 Dominante Frequenzen der Freese-Reihe:", np.round(freq_dom_freese, 6))
print("🔥 Dominante Amplituden der Freese-Reihe:", np.round(amp_dom_freese, 2))


# 📌 Plot: Fourier-Spektrum der drei Funktionen
plt.figure(figsize=(10, 5))
plt.plot(freq_zeta.cpu().numpy(), torch.abs(fft_zeta).cpu().numpy(), label="Zeta-Nullstellen", color="b")
plt.plot(freq_beta.cpu().numpy(), torch.abs(fft_beta).cpu().numpy(), label="Beta-Skala", color="g")
plt.plot(freq_freese.cpu().numpy(), torch.abs(fft_freese).cpu().numpy(), label="Freese-Reihe", color="r")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektren der Zeta-Nullstellen, Beta-Skala und Freese-Reihe")
plt.legend()
plt.grid()
plt.show()
```
