```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# === Parameter ===
dateipfad = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"  # <== hier anpassen!
anzahl_cluster = 5

# === Daten laden ===
betas = np.loadtxt(dateipfad)

# === Punkte auf dem Einheitskreis berechnen ===
z = np.exp(1j * np.pi * betas)
punkte = np.column_stack((z.real, z.imag))

# === Clustering ===
kmeans = KMeans(n_clusters=anzahl_cluster, n_init='auto').fit(punkte)
labels = kmeans.labels_
zentren = kmeans.cluster_centers_

# === Winkel in Grad berechnen ===
winkel = np.degrees(np.arctan2(zentren[:,1], zentren[:,0]))
winkel = np.mod(winkel + 180, 360) - 180  # in [−180°, 180°]

# === Ergebnis anzeigen ===
print("Dominante Winkelzentren (in Grad):")
for i, w in enumerate(winkel):
    print(f"Cluster {i+1}: {w:.2f}°")

# === Plot ===
plt.figure(figsize=(6,6))
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.plot(np.cos(np.linspace(0, 2*np.pi, 300)),
         np.sin(np.linspace(0, 2*np.pi, 300)), 'k--', alpha=0.3, label="Einheitskreis")
plt.scatter(punkte[:,0], punkte[:,1], c=labels, cmap='Set2', s=15, alpha=0.7)
plt.scatter(zentren[:,0], zentren[:,1], c='red', s=100, label='Clusterzentren')
plt.xlabel("Reelle Achse")
plt.ylabel("Imaginäre Achse")
plt.title(r"$e^{i \beta \pi}$ auf dem Einheitskreis (Clustering)")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# === Parameter ===
dateipfad = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv" # <== anpassen!
anzahl_cluster = 5

# === Daten laden (mit Header) ===
daten = np.genfromtxt(dateipfad, delimiter=",", skip_header=1)
betas = daten[:, 0] if daten.ndim > 1 else daten  # Falls nur eine Spalte

# === Punkte auf dem Einheitskreis berechnen ===
z = np.exp(1j * np.pi * betas)
punkte = np.column_stack((z.real, z.imag))

# === Clustering ===
kmeans = KMeans(n_clusters=anzahl_cluster, n_init='auto').fit(punkte)
labels = kmeans.labels_
zentren = kmeans.cluster_centers_

# === Winkel in Grad berechnen ===
winkel = np.degrees(np.arctan2(zentren[:,1], zentren[:,0]))
winkel = np.mod(winkel + 180, 360) - 180  # in [−180°, 180°]

# === Ergebnis anzeigen ===
print("Dominante Winkelzentren (in Grad):")
for i, w in enumerate(winkel):
    print(f"Cluster {i+1}: {w:.2f}°")

# === Plot ===
plt.figure(figsize=(6,6))
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)
plt.plot(np.cos(np.linspace(0, 2*np.pi, 300)),
         np.sin(np.linspace(0, 2*np.pi, 300)), 'k--', alpha=0.3, label="Einheitskreis")
plt.scatter(punkte[:,0], punkte[:,1], c=labels, cmap='Set2', s=15, alpha=0.7)
plt.scatter(zentren[:,0], zentren[:,1], c='red', s=100, label='Clusterzentren')
plt.xlabel("Reelle Achse")
plt.ylabel("Imaginäre Achse")
plt.title(r"$e^{i \beta \pi}$ auf dem Einheitskreis (Clustering)")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()
```
