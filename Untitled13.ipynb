```python
# ‚úÖ 1. Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import requests
from scipy.optimize import curve_fit

# ‚úÖ 2. Online-Datenquellen definieren
urls = {
    "100 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "100.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2.000.000 Nullstellen": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "50.000 Nullstellen (LMFDB)": "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1",
    "100.000 Nullstellen (LMFDB)": "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"
}

# ‚úÖ 3. Funktion zum Laden und Bereinigen der Daten
def load_nullstellen(url, is_lmfdb=False):
    """ L√§dt Nullstellen aus einer Online-Quelle """
    response = requests.get(url)
    if response.status_code == 200:
        lines = response.text.splitlines()
        if is_lmfdb:
            data = np.array([float(line.split()[1]) for line in lines if line.strip() and line[0].isdigit()])
        else:
            data = np.array([float(line.split()[0]) for line in lines if line.strip()])
        print(f"‚úÖ {len(data)} Nullstellen erfolgreich aus {url} geladen!")
        return data
    else:
        print(f"‚ùå Fehler beim Laden der Daten von {url}")
        return None

# ‚úÖ 4. Alle Daten laden
nullstellen_daten = {}
for name, url in urls.items():
    is_lmfdb = "lmfdb" in url
    data = load_nullstellen(url, is_lmfdb)
    if data is not None:
        nullstellen_daten[name] = data

# ‚úÖ 5. Funktion zur Berechnung der Abst√§nde
def berechne_abstaende(nullstellen):
    return np.diff(nullstellen)

# ‚úÖ 6. Berechnung der Nullstellen-Abst√§nde
abstaende_daten = {name: berechne_abstaende(data) for name, data in nullstellen_daten.items()}

# ‚úÖ 7. Freese-Formel definieren
def freese_formula(N, A, beta):
    return A * N ** beta

# ‚úÖ 8. Curve-Fitting der Freese-Formel auf alle Datens√§tze
fit_ergebnisse = {}

for name, abstaende in abstaende_daten.items():
    N_values = np.arange(1, len(abstaende) + 1)
    try:
        params, _ = curve_fit(freese_formula, N_values, abstaende, p0=[1.5, 0.48])
        fit_ergebnisse[name] = params
        print(f"\nüìä Optimierte Parameter f√ºr {name}:")
        print(f"üîπ Skalierungsfaktor A = {params[0]:.6f}")
        print(f"üîπ Experimenteller Exponent beta = {params[1]:.6f} (Erwartet: 0.484964)")
    except Exception as e:
        print(f"‚ö†Ô∏è Fehler beim Curve-Fit f√ºr {name}: {e}")

# ‚úÖ 9. Visualisierung der Fits
plt.figure(figsize=(12, 6))

for name, abstaende in abstaende_daten.items():
    if name in fit_ergebnisse:
        N_values = np.arange(1, len(abstaende) + 1)
        plt.loglog(N_values, abstaende, 'o', markersize=2, alpha=0.5, label=f"{name} Daten")
        plt.loglog(N_values, freese_formula(N_values, *fit_ergebnisse[name]), '--', label=f"Fit {name}")

plt.xlabel("Index N")
plt.ylabel("Nullstellen-Abst√§nde")
plt.legend()
plt.title("Vergleich: Freese-Formel f√ºr verschiedene Nullstellen-Datens√§tze")
plt.grid(True)
plt.show()
```

```python
# ‚úÖ Importiere die ben√∂tigten Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Lade Nullstellen-Daten (Echte Daten oder Simulation)
def lade_nullstellen_txt(dateipfad):
    try:
        with open(dateipfad, "r") as file:
            nullstellen = np.array([float(line.strip()) for line in file])
        abstaende = np.diff(nullstellen)  # Abstand zwischen Nullstellen berechnen
        return np.arange(1, len(abstaende) + 1), abstaende
    except Exception as e:
        print(f"‚ùå Fehler beim Laden der Nullstellen aus {dateipfad}: {e}")
        return None, None

# ‚úÖ Fibonacci-Freese-Formel mit log und Sinus-Korrektur
def fibonacci_freese_extended(N, A, beta, C, B, omega, phi):
    return A * N**beta + C * np.log(N) + B * np.sin(omega * N + phi)

# ‚úÖ Initiale Sch√§tzung der Parameter
initial_params = [1.5, 0.48, 0.1, 0.01, 0.01, 0]  # (A, beta, C, B, omega, phi)

# ‚úÖ Pfade zu den echten Nullstellen-Dateien
pfad_2M = "/content/drive/MyDrive/zeros6.txt"
pfad_3M = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros_30607946000.dat"  # Muss mit platt_zeros.py geladen werden

# ‚úÖ Lade die 2M Nullstellen (Echte Daten)
N_2M, abstaende_2M = lade_nullstellen_txt(pfad_2M)

# ‚úÖ Simuliere Daten f√ºr 3M falls Platt-Import fehlschl√§gt
N_3M = np.linspace(1, 3_000_000, num=100)  # Simulierte N-Werte
abstaende_3M = 1.5 * N_3M**0.48 + 0.05 * np.log(N_3M) + 0.02 * np.sin(0.01 * N_3M)

# ‚úÖ Curve-Fit f√ºr 2M Nullstellen
if N_2M is not None:
    params_2M, _ = curve_fit(fibonacci_freese_extended, N_2M, abstaende_2M, p0=initial_params)
    A_2M, beta_2M, C_2M, B_2M, omega_2M, phi_2M = params_2M

    # ‚úÖ Ergebnisse f√ºr 2M ausgeben
    print("\nüìä **Optimierte Parameter f√ºr 2M-Nullstellen:**")
    print(f"üîπ Skalierungsfaktor A = {A_2M:.6f}")
    print(f"üîπ Experimenteller Exponent beta = {beta_2M:.6f} (Erwartet: 0.484964)")
    print(f"üîπ Logarithmischer Korrekturterm C = {C_2M:.6f}")
    print(f"üîπ Sinus-Amplitude B = {B_2M:.6f}")
    print(f"üîπ Sinus-Frequenz omega = {omega_2M:.6f}")
    print(f"üîπ Sinus-Phase phi = {phi_2M:.6f}")

# ‚úÖ Curve-Fit f√ºr 3M Nullstellen (Simulation)
params_3M, _ = curve_fit(fibonacci_freese_extended, N_3M, abstaende_3M, p0=initial_params)
A_3M, beta_3M, C_3M, B_3M, omega_3M, phi_3M = params_3M

# ‚úÖ Ergebnisse f√ºr 3M ausgeben
print("\nüìä **Optimierte Parameter f√ºr 3M-Nullstellen (Simulation):**")
print(f"üîπ Skalierungsfaktor A = {A_3M:.6f}")
print(f"üîπ Experimenteller Exponent beta = {beta_3M:.6f} (Erwartet: 0.484964)")
print(f"üîπ Logarithmischer Korrekturterm C = {C_3M:.6f}")
print(f"üîπ Sinus-Amplitude B = {B_3M:.6f}")
print(f"üîπ Sinus-Frequenz omega = {omega_3M:.6f}")
print(f"üîπ Sinus-Phase phi = {phi_3M:.6f}")

# ‚úÖ Visualisierung f√ºr beide
plt.figure(figsize=(10, 5))
plt.scatter(N_2M, abstaende_2M, label="Echte 2M Daten", color="blue", alpha=0.5)
plt.plot(N_2M, fibonacci_freese_extended(N_2M, *params_2M), label="Fit f√ºr 2M Daten", color="red", linewidth=2)

plt.scatter(N_3M, abstaende_3M, label="Simulierte 3M Daten", color="green", alpha=0.5)
plt.plot(N_3M, fibonacci_freese_extended(N_3M, *params_3M), label="Fit f√ºr 3M Daten", color="orange", linewidth=2)

plt.xscale("log")
plt.yscale("log")
plt.xlabel("N (Index der Nullstelle)")
plt.ylabel("Koh√§renzl√§nge L(N)")
plt.title("Vergleich: Echte 2M vs. Simulierte 3M Daten")
plt.legend()
plt.grid()
plt.show()
```
