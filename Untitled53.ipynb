```python
# --- Setup ---
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.optimize import curve_fit
from scipy.special import ellipj

# --- Hauptfunktion: FFF ohne Korrekturterm ---
def FFF_main(N, A=1.0, beta=0.48):
    return A * N**beta

# --- Funktion mit log-Korrekturterm (Fourier-Ansatz) ---
def FFF_logcorr(N, A, beta, C, omega):
    return A * N**beta * (1 + C * np.cos(omega * np.log(N)))

# --- Funktion mit elliptischem Korrekturterm (Jacobi cn) ---
def FFF_cn(N, A, beta, C, omega, m):
    cn_vals = ellipj(omega * np.log(N), m)[1]  # cn-Komponente
    return A * N**beta * (1 + C * cn_vals)

# --- Simulation einer modulierteren Wahrheit (z.B. durch Zeta-Nullstellen induziert) ---
def true_function(N):
    A, beta, C, omega = 1.0, 0.48, 0.1, 3.0
    return FFF_logcorr(N, A, beta, C, omega)

# --- Daten vorbereiten ---
N = np.linspace(10, 1000, 500)
y_true = true_function(N)
y_main = FFF_main(N)

# --- Residuum & FFT ---
residuum = y_true - y_main
res_fft = fft(residuum)
freqs = fftfreq(len(N), d=(N[1]-N[0]))
amplitudes = np.abs(res_fft)

# --- Dominante Frequenz finden ---
pos_freqs = freqs[:len(freqs)//2]
pos_amp = amplitudes[:len(amplitudes)//2]
dominant_idx = np.argmax(pos_amp[1:]) + 1
omega_guess = 2 * np.pi * pos_freqs[dominant_idx]
print(f">>> Dominante Frequenz erkannt: œâ ‚âà {omega_guess:.4f}")

# --- Fit 1: Kosinus-Korrekturterm ---
popt_log, _ = curve_fit(
    FFF_logcorr, N, y_true, p0=[1.0, 0.48, 0.1, omega_guess]
)

# --- Fit 2: Jacobi cn (elliptisch) ---
popt_cn, _ = curve_fit(
    FFF_cn, N, y_true, p0=[1.0, 0.48, 0.1, omega_guess, 0.9]
)

# --- Visualisierung ---
plt.figure(figsize=(10,6))
plt.plot(N, y_true, label="Wahr (z.B. FFT mit Mod.)", linewidth=2)
plt.plot(N, FFF_main(N), '--', label="FFF (ohne Korrektur)")
plt.plot(N, FFF_logcorr(N, *popt_log), '--', label="Fit: log-Kosinus")
plt.plot(N, FFF_cn(N, *popt_cn), '--', label="Fit: Jacobi-cn")

plt.title("Korrektur-Fits auf FFF")
plt.xlabel("N")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Parametervergleich ---
print(">>> Fit-Ergebnisse (log-kosinus):")
print(f"A = {popt_log[0]:.4f}, Œ≤ = {popt_log[1]:.4f}, C = {popt_log[2]:.4f}, œâ = {popt_log[3]:.4f}")
print(">>> Fit-Ergebnisse (Jacobi-cn):")
print(f"A = {popt_cn[0]:.4f}, Œ≤ = {popt_cn[1]:.4f}, C = {popt_cn[2]:.4f}, œâ = {popt_cn[3]:.4f}, m = {popt_cn[4]:.4f}")
```

```python
# üìå 1Ô∏è‚É£ Notwendige Bibliotheken installieren
!pip install gdown numpy pandas matplotlib scipy

# üìå 2Ô∏è‚É£ Bibliotheken importieren
import gdown
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# üìå 3Ô∏è‚É£ **Daten-Download von Google Drive**
data_urls = {
    "zeta_nullstellen": "https://drive.google.com/uc?id=1P0Bb_-Gu1Bc_WWVEMabZBjvMfzEo46NB",
    "beta_skala": "https://drive.google.com/uc?id=1Zkz8ycT_HnHzP_rUsPtLlU93mu1qyICA",
    "freese_funktion": "https://drive.google.com/uc?id=1ZL0uigQDakQomhD6bjKZPRjyae8x6v29"
}

data_files = {}
for name, url in data_urls.items():
    output = f"/content/{name}.csv"
    gdown.download(url, output, quiet=False)
    data_files[name] = output

# üìå 4Ô∏è‚É£ **Daten einlesen**
data = {}
for name, file in data_files.items():
    try:
        df = pd.read_csv(file, header=None).dropna()
        df = df.applymap(lambda x: pd.to_numeric(str(x).replace(",", "."), errors='coerce'))  # Komma zu Punkt
        data[name] = df[0].values
        print(f"‚úÖ {name}: {len(data[name])} Werte geladen!")
    except Exception as e:
        print(f"‚ö†Ô∏è Fehler beim Einlesen von {name}: {e}")

# üìå 5Ô∏è‚É£ **Freese-Funktionen definieren**
def freese_original(n, A=0.47, B=0.50, C=0.02):
    return A * n**B + C

def freese_fibonacci(n, A=0.47, B=0.50, C=0.02, phi=1.618):
    return A * n**B * np.exp(phi/n)

def freese_operator(n, A=0.47, B=0.50, C=0.02):
    return A * n**B + C * np.log(n + 1)

# üìå 6Ô∏è‚É£ **Datenbereich definieren**
N = 1000
n_values = np.arange(1, N+1)

# üìå 7Ô∏è‚É£ **Funktionwerte berechnen**
freese_values = {
    "Original": freese_original(n_values),
    "Fibonacci": freese_fibonacci(n_values),
    "Operator": freese_operator(n_values)
}

# üìå 8Ô∏è‚É£ **Vergleich der Freese-Formeln plotten**
plt.figure(figsize=(8, 5))
for label, values in freese_values.items():
    plt.plot(n_values, values, label=label)
plt.xlabel("n")
plt.ylabel("L(n)")
plt.title("Vergleich der Freese-Formeln")
plt.legend()
plt.grid()
plt.show()

# üìå 9Ô∏è‚É£ **Fourier-Analyse f√ºr Beta-Skala & Freese-Funktion**
for key in ["beta_skala", "freese_funktion"]:
    if key in data:
        values = data[key][:1000]  # Beschr√§nken auf 1000 Werte
        freq_spectrum = np.abs(fft(values))
        freqs = fftfreq(len(values))

        plt.figure(figsize=(7, 4))
        plt.plot(freqs[:len(freqs)//2], freq_spectrum[:len(freqs)//2], label=key, color="red" if key == "freese_funktion" else "blue")
        plt.xlabel("Frequenz")
        plt.ylabel("Spektrale Intensit√§t")
        plt.title(f"Fourier-Spektrum der {key.replace('_', ' ').title()}")
        plt.legend()
        plt.grid()
        plt.show()

# üìå üîü **Korrelationen mit den Zeta-Nullstellen berechnen**
correlations = {}
for label, values in freese_values.items():
    min_length = min(len(values), len(data["zeta_nullstellen"]))
    correlations[f"{label} vs Zeta-Nullstellen"] = np.corrcoef(values[:min_length], data["zeta_nullstellen"][:min_length])[0, 1]

# üìå üîü **Ergebnisse ausgeben**
print("\nüìä **Korrelationen mit Zeta-Nullstellen:**")
for key, value in correlations.items():
    print(f"{key}: {value:.5f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üìå 1Ô∏è‚É£ Datenpfad f√ºr Odlyzko-Zeta-Nullstellen
zeta_zeros_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.csv"

# üìå 2Ô∏è‚É£ Daten einlesen & bereinigen
def load_clean_csv(file_path, max_values=2_000_000):
    df = pd.read_csv(file_path, header=None, skiprows=1)  # Header √ºberspringen
    df = df.apply(pd.to_numeric, errors='coerce').dropna()  # Strings in Zahlen umwandeln & NaNs entfernen
    return df.iloc[:max_values, :].values.flatten()  # Daten als 1D-Array zur√ºckgeben

# üìå 3Ô∏è‚É£ Daten laden
zeta_zeros = load_clean_csv(zeta_zeros_path)

# üìå 4Ô∏è‚É£ Begrenzung auf 2 Mio Nullstellen
N = len(zeta_zeros)
N_vals = np.arange(2, N+1)  # Start bei 2, um log(0) zu vermeiden

print(f"‚úÖ Zeta-Nullstellen geladen: {len(zeta_zeros)} Werte")

# üî¥ Modell 1: Logarithmische Approximation
def model_1(N, A, B):
    return A - B * np.log(N)

# üîµ Modell 2: Euler-Freese-Korrektur
def model_2(N, A, B):
    zeta_correction = np.array([(-1)**n * (1 / (n+1)**2) for n in range(1, len(N)+1)])
    return A * (1 / (np.log(N + 1e-10))) + B * zeta_correction

# üü¢ Modell 3: Fibonacci-korrigierte Form
phi = (1 + np.sqrt(5)) / 2  # Goldener Schnitt
def model_3(N, A, B, C):
    return A - B * np.log(N) + C * np.exp(- (1 / (phi**2 * np.pi)) - (np.exp(-phi) / (np.pi**2)) - (np.log(N) / (phi**3 * np.pi**3)))

# üìå 5Ô∏è‚É£ Curve-Fitting mit den echten Nullstellen
popt_1, _ = curve_fit(model_1, N_vals, zeta_zeros[1:], maxfev=10000)
popt_2, _ = curve_fit(model_2, N_vals, zeta_zeros[1:], maxfev=10000)
popt_3, _ = curve_fit(model_3, N_vals, zeta_zeros[1:], maxfev=10000)

# üìå 6Ô∏è‚É£ Berechnung von R¬≤ f√ºr jedes Modell
def r_squared(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

r2_1 = r_squared(zeta_zeros[1:], model_1(N_vals, *popt_1))
r2_2 = r_squared(zeta_zeros[1:], model_2(N_vals, *popt_2))
r2_3 = r_squared(zeta_zeros[1:], model_3(N_vals, *popt_3))

print(f"üìä R¬≤-Werte:")
print(f"   üî¥ Log-Modell: R¬≤ = {r2_1:.6f}")
print(f"   üîµ Euler-Freese-Modell: R¬≤ = {r2_2:.6f}")
print(f"   üü¢ Fibonacci-Modell: R¬≤ = {r2_3:.6f}")

# üìå 7Ô∏è‚É£ Visualisierung
plt.figure(figsize=(7,5))
plt.plot(N_vals, zeta_zeros[1:], 'k.', alpha=0.5, label="Echte Zeta-Nullstellen")
plt.plot(N_vals, model_1(N_vals, *popt_1), 'r-', label=f"Log-Modell: {popt_1[0]:.4f} - {popt_1[1]:.4f} log(N), R¬≤={r2_1:.6f}")
plt.plot(N_vals, model_2(N_vals, *popt_2), 'b--', label=f"Euler-Freese-Modell, R¬≤={r2_2:.6f}")
plt.plot(N_vals, model_3(N_vals, *popt_3), 'g-.', label=f"Fibonacci-Modell, R¬≤={r2_3:.6f}")
plt.xlabel("Index N")
plt.ylabel("Zeta-Nullstellenwerte")
plt.title("Vergleich: Modelle vs. echte Zeta-Nullstellen")
plt.legend()
plt.grid()
plt.show()
```
