```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Datei laden (anpassen falls nötig) ===
file_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
zeta_zeros_full = pd.read_csv(file_path, header=None).squeeze()
zeta_zeros_full = pd.to_numeric(zeta_zeros_full, errors='coerce').dropna().values

# === Für erste Analyse z.B. 5000 Nullstellen verwenden ===
N = 5000
zeta_zeros = zeta_zeros_full[:N]
n_vals = np.arange(1, N + 1)

# === Deine ZNF-Formel ===
def znf(n, A, beta, C):
    return A * n**beta + C * np.log(n)

# === Parameter fitten ===
params, _ = curve_fit(znf, n_vals, zeta_zeros, p0=[2.0, 0.9, 0.1])
A, beta, C = params

# === Rekonstruktion & Fehleranalyse ===
znf_fit = znf(n_vals, A, beta, C)
delta = znf_fit - zeta_zeros
corr = np.corrcoef(zeta_zeros, znf_fit)[0, 1]
mse = np.mean(delta**2)

# === Ergebnisse anzeigen ===
print(f"\n✅ Fit-ZNF-Parameter: A = {A:.6f}, β = {beta:.6f}, C = {C:.6f}")
print(f"📊 Pearson-Korrelation: {corr:.8f}")
print(f"📉 MSE: {mse:.8f}")

# === Plot 1: Vergleich ===
plt.figure(figsize=(10, 5))
plt.plot(n_vals, zeta_zeros, label="Zeta-Nullstellen", linewidth=1.5)
plt.plot(n_vals, znf_fit, '--', label="ZNF-Fit", linewidth=1.5)
plt.title("Rekonstruktion der RH-Nullstellen aus ZNF-Formel")
plt.xlabel("Index n")
plt.ylabel("Imaginärteil ρ")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# === Plot 2: Fehlerverlauf ===
plt.figure(figsize=(10, 3))
plt.plot(n_vals, delta, color='purple')
plt.axhline(0, linestyle='--', color='gray')
plt.title("Abweichung: ZNF - Zeta")
plt.xlabel("Index n")
plt.ylabel("Fehler")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# === Bibliotheken laden ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, fftfreq

# === Datei laden ===
path = "/content/drive/MyDrive/beta_skala_praezision_cleaned.csv"
df = pd.read_csv(path, header=None)
beta_raw = pd.to_numeric(df[0], errors='coerce').dropna().values
beta = beta_raw[:10000]  # anpassen bei Bedarf

# === Kohärenzlängen berechnen: L(n) = β(n+1) - β(n)
L = np.diff(beta)
n_vals = np.arange(1, len(L) + 1)

# === Modellfunktion: Gesetz L(n) = A * n^β + C * sin(ω * log(n))
def L_model(n, A, beta, C, ω):
    return A * n**beta + C * np.sin(ω * np.log(n))

# === Curve Fit
popt, _ = curve_fit(L_model, n_vals, L, p0=[0.1, -0.5, 0.01, 5.0])
A_fit, beta_fit, C_fit, omega_fit = popt

# === Modell aus Fit erzeugen
L_fit = L_model(n_vals, *popt)

# === Plot: Rohdaten & Fit
plt.figure(figsize=(10, 5))
plt.plot(n_vals, L, '.', label="Kohärenzlängen Δβ(n)", alpha=0.4)
plt.plot(n_vals, L_fit, 'r-', label=f"Fit: $L(n) = A n^{{β}} + C \\sin(ω \\log n)$")
plt.xlabel("n")
plt.ylabel("L(n)")
plt.title("Kohärenzlängen-Gesetz Δβ(n)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# === Fehler und Statistik
mse = np.mean((L - L_fit)**2)
correlation = np.corrcoef(L, L_fit)[0, 1]
print(f"✅ Angepasste Parameter:")
print(f"A = {A_fit:.6f}, β = {beta_fit:.6f}, C = {C_fit:.6f}, ω = {omega_fit:.6f}")
print(f"📉 MSE: {mse:.10f}")
print(f"📊 Pearson-Korrelation: {correlation:.8f}")

# === Fourier-Analyse: Frequenzkomponenten von L(n)
fft_vals = fft(L - np.mean(L))
freqs = fftfreq(len(L), d=1)

plt.figure(figsize=(10, 4))
plt.plot(freqs[:len(freqs)//2], np.abs(fft_vals[:len(freqs)//2]), 'purple')
plt.title("Spektrum der Kohärenzlängen Δβ(n)")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Parameter
N = 10000  # Anzahl Nullstellen
A, beta, C, eps = 3.821226, 0.852602, 9.132141, 2.396560  # ZNF-Operator-Parameter

# === Zeta-Modell (synthetische Nullstellen)
n_vals = np.arange(1, N+1)
zeta_n = A * n_vals**beta + C * np.log(n_vals) + eps * np.random.normal(0, 1, N)

# === Konstruktion des Operator-Faktors e^{iπβ}
def compute_beta_operator(zeta):
    beta_out = np.zeros(len(zeta))
    for i in range(1, len(zeta)-1):
        delta1 = zeta[i] - zeta[i-1]
        delta2 = zeta[i+1] - zeta[i]
        # Schätzung basierend auf lokalen Differenzen
        beta_out[i] = 1 if delta2 > delta1 else 0
    beta_out[0] = beta_out[-1] = 1  # Ränder sichern
    return beta_out

beta_gold = compute_beta_operator(zeta_n)

# === Speichern als CSV
df_beta = pd.DataFrame(beta_gold, columns=["beta"])
df_beta.to_csv("/content/beta_goldstandard_operator.csv", index=False)
print("✅ CSV gespeichert als: beta_goldstandard_operator.csv")

# === Plot
plt.figure(figsize=(12, 4))
plt.plot(n_vals, beta_gold, color="orange", linewidth=2)
plt.title("Goldstandard-Beta-Skala aus Operator")
plt.xlabel("Index n")
plt.ylabel(r"$\beta(n)$")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# === Parameter
N = 10_000
A, beta, C, eps = 2.5, 0.88, 4.0, 1.2

n_vals = np.arange(1, N + 1)
eps_term = eps * np.sin(0.01 * n_vals)  # Langsame Modulation
zeta_n = A * n_vals**beta + C * np.log(n_vals) + eps_term

# === Verbesserter Operator
def beta_operator_precise(zeta, window=5):
    beta_out = np.zeros(len(zeta))
    for i in range(window, len(zeta) - window):
        local_prev = np.mean(np.diff(zeta[i - window:i]))
        local_next = np.mean(np.diff(zeta[i:i + window]))
        ratio = abs(local_next / local_prev) if local_prev != 0 else 1.0
        beta_out[i] = 1 if abs(ratio - 1) > 0.01 else 0  # Empfindlicher Schwellenwert
    beta_out[:window] = beta_out[-window:] = 1
    return beta_out

# === Neue Beta-Skala berechnen
beta_values = beta_operator_precise(zeta_n)

# === Plot
plt.figure(figsize=(12, 4))
plt.plot(n_vals, beta_values, color='orange', lw=2)
plt.title("Goldstandard-Beta-Skala aus Operator (sensibel)")
plt.xlabel("Index n")
plt.ylabel(r"$\beta(n)$")
plt.grid(True)
plt.show()

# === CSV speichern
pd.DataFrame(beta_values).to_csv("goldstandard_beta_operator_sensibel.csv", index=False, header=False)
print("✅ Datei gespeichert: goldstandard_beta_operator_sensibel.csv")
```

```python
# === Bibliotheken ===
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit, root_scalar
from scipy.special import zeta as zeta_special
from math import log

# === Dirichlet-Zeichen mod 8 (quadratisch für Q(sqrt(2))) ===
def chi_mod8(n):
    r = n % 8
    if r in (1, 7): return 1
    elif r in (3, 5): return -1
    else: return 0

# === Dedekind-Zeta-Funktion: ζ(s) · L(s, χ) ===
def dedekind_zeta(s, N=1000):
    try:
        L_s = sum(chi_mod8(n) / (n ** s) for n in range(1, N) if chi_mod8(n) != 0)
        return zeta_special(s.real) * L_s  # Imaginärteil von ζ wird nicht unterstützt
    except:
        return 1e6  # Fehlerfall behandeln

# === Realteil entlang kritischer Linie: Re(ζ_K(0.5 + it)) ===
def real_part(t):
    s = 0.5 + 1j * t
    try:
        return dedekind_zeta(s).real
    except:
        return 1e6

# === Gitterprobe: t ∈ [10, 200] ===
t_vals = np.linspace(10, 200, 10000)
vals = np.array([real_part(t) for t in t_vals])

# === Lokale Minima als Nullstellenkandidaten (robust!) ===
approx_zeros = []
threshold = 0.5
for i in range(1, len(vals) - 1):
    if vals[i-1] > vals[i] < vals[i+1] and abs(vals[i]) < threshold:
        approx_zeros.append(t_vals[i])

# === Präzisieren mit Brent-Solver ===
dedekind_zeros = []
for t0 in approx_zeros[:50]:  # Weniger für mehr Stabilität
    try:
        res = root_scalar(real_part, bracket=[t0 - 0.3, t0 + 0.3], method='brentq')
        if res.converged:
            dedekind_zeros.append(res.root)
    except:
        continue

print(f"✅ Nullstellen gefunden: {len(dedekind_zeros)}")

# === ZNF-Modell: Fit gegen A·n^β + C·log(n) ===
def znf_model(n, A, beta, C):
    return A * n**beta + C * np.log(n)

n_vals = np.arange(1, len(dedekind_zeros)+1)
zeros_arr = np.array(dedekind_zeros)

# === Curve Fit durchführen
popt, _ = curve_fit(znf_model, n_vals, zeros_arr, p0=[3.0, 0.9, 5.0])
A, beta, C = popt
znf_fit = znf_model(n_vals, A, beta, C)

# === Statistische Auswertung
corr = np.corrcoef(zeros_arr, znf_fit)[0, 1]
mse = np.mean((znf_fit - zeros_arr)**2)

print(f"\n✅ Angepasste ZNF-Parameter:")
print(f"A = {A:.6f},  β = {beta:.6f},  C = {C:.6f}")
print(f"📊 Pearson-Korrelation: {corr:.8f}")
print(f"📉 MSE: {mse:.8f}")

# === Plot
plt.figure(figsize=(8, 5))
plt.plot(n_vals, zeros_arr, label="Dedekind-Zeta Nullstellen", linewidth=2)
plt.plot(n_vals, znf_fit, '--', label="ZNF-Fit", linewidth=2)
plt.xlabel("Index n")
plt.ylabel("Imaginärteil")
plt.title("Dedekind-Zeta Nullstellen vs. ZNF-Modell")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# === Bibliotheken laden ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, fftfreq

# === Datei laden ===
path = "/content/drive/MyDrive/zeros1.txt"
df = pd.read_csv(path, header=None)
beta_raw = pd.to_numeric(df[0], errors='coerce').dropna().values
beta = beta_raw[:10000]  # anpassen bei Bedarf

# === Kohärenzlängen berechnen: L(n) = β(n+1) - β(n)
L = np.diff(beta)
n_vals = np.arange(1, len(L) + 1)

# === Modellfunktion: Gesetz L(n) = A * n^β + C * sin(ω * log(n))
def L_model(n, A, beta, C, ω):
    return A * n**beta + C * np.sin(ω * np.log(n))

# === Curve Fit
popt, _ = curve_fit(L_model, n_vals, L, p0=[0.1, -0.5, 0.01, 5.0])
A_fit, beta_fit, C_fit, omega_fit = popt

# === Modell aus Fit erzeugen
L_fit = L_model(n_vals, *popt)

# === Plot: Rohdaten & Fit
plt.figure(figsize=(10, 5))
plt.plot(n_vals, L, '.', label="Kohärenzlängen Δβ(n)", alpha=0.4)
plt.plot(n_vals, L_fit, 'r-', label=f"Fit: $L(n) = A n^{{β}} + C \\sin(ω \\log n)$")
plt.xlabel("n")
plt.ylabel("L(n)")
plt.title("Kohärenzlängen-Gesetz Δβ(n)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# === Fehler und Statistik
mse = np.mean((L - L_fit)**2)
correlation = np.corrcoef(L, L_fit)[0, 1]
print(f"✅ Angepasste Parameter:")
print(f"A = {A_fit:.6f}, β = {beta_fit:.6f}, C = {C_fit:.6f}, ω = {omega_fit:.6f}")
print(f"📉 MSE: {mse:.10f}")
print(f"📊 Pearson-Korrelation: {correlation:.8f}")

# === Fourier-Analyse: Frequenzkomponenten von L(n)
fft_vals = fft(L - np.mean(L))
freqs = fftfreq(len(L), d=1)

plt.figure(figsize=(10, 4))
plt.plot(freqs[:len(freqs)//2], np.abs(fft_vals[:len(freqs)//2]), 'purple')
plt.title("Spektrum der Kohärenzlängen Δβ(n)")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Beta-Skala laden (analytisch)
path = "/content/beta_spektrum_analytisch.csv"
beta = pd.read_csv(path, header=None).squeeze().dropna().astype(float).values
n_vals = np.arange(1, len(beta) + 1)

# === Fehlerterm berechnen
delta = 0.5 - beta

# === Modell: δ(n) = c / log(n)^a
def delta_model(n, c, a):
    return c / (np.log(n)**a)

# === Curve Fit durchführen
popt, _ = curve_fit(delta_model, n_vals, delta, p0=[0.01, 1.0])
c_fit, a_fit = popt
delta_fit = delta_model(n_vals, *popt)

# === Plot
plt.figure(figsize=(10, 5))
plt.plot(n_vals, delta, '.', label="δ(n) = 0.5 - β(n)", alpha=0.5)
plt.plot(n_vals, delta_fit, 'r-', label=f"Fit: δ(n) = {c_fit:.6f} / log(n)^{a_fit:.3f}")
plt.xlabel("n")
plt.ylabel("δ(n)")
plt.title("Struktur des Fehlerterms δ(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === Statistik
mse = np.mean((delta - delta_fit)**2)
correlation = np.corrcoef(delta, delta_fit)[0, 1]

print(f"✅ Angepasste Parameter: c = {c_fit:.10f}, a = {a_fit:.10f}")
print(f"📉 MSE: {mse:.10f}")
print(f"📊 Pearson-Korrelation: {correlation:.8f}")
```

```python
# === Bibliotheken laden ===
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from numpy import pi

# === Parameter ===
N = 10000  # Anzahl der Nullstellen / Operatorgröße
A, beta, C, epsilon = 3.157144, 0.872137, 16.093850, 0.256256

# === ZNF-Modell: Beta-Skala (rekonstruiert)
n_vals = np.arange(1, N + 1)
beta_n = A * n_vals**beta + C * np.log(n_vals) + epsilon * np.sin(n_vals)

# === Operator: Hamilton H als symmetrische Matrix
main_diag = beta_n
off_diag = np.ones(N - 1)
H = np.diag(main_diag) + np.diag(off_diag, 1) + np.diag(off_diag, -1)

# === Modulierter Operator mit Phase: H * exp(iπβ_n)
phase = np.exp(1j * pi * beta_n)
H_mod = H.astype(complex) @ np.diag(phase)

# === Eigenwerte berechnen
eigvals = np.linalg.eigvals(H_mod)

# === Plot: Komplexes Spektrum (Spirale) ===
plt.figure(figsize=(7, 7))
plt.scatter(eigvals.real, eigvals.imag, color='purple', s=4)
plt.title(r"Komplexe Eigenwerte von $H \cdot e^{i\pi\beta_n}$")
plt.xlabel("Realteil")
plt.ylabel("Imaginärteil")
plt.grid(True)
plt.axis("equal")
plt.tight_layout()
plt.show()
```
