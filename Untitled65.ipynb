```python
# === Bibliotheken laden ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# === Dateipfade ===
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'
beta_path = '/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv'

# === Daten laden ===
gamma = pd.read_csv(zeta_path, header=None).iloc[:, 0].astype(float).to_numpy()
primes = pd.read_csv(prime_path, skiprows=1, header=None).iloc[:, 0].astype(int).to_numpy()
beta = pd.read_csv(beta_path, header=None).iloc[:, 0].astype(float).to_numpy()

# === Echte ψ(x) berechnen (Chebyshev-ψ)
psi_true = np.cumsum(np.log(primes))

# === Bereich für Modell
N_vals = np.arange(1, len(primes[:len(gamma)]) + 1)
x_vals = primes[:len(N_vals)]

# === FFF-Modell (rekonstruktiv mit β(n))
def psi_model(x, gamma, beta):
    rho = 0.5 + 1j * gamma
    xlog = np.log(x + 1e-10)
    terms = beta * np.power(x[:, None], rho) / (rho * (xlog[:, None] + 1))
    return np.real(np.sum(terms, axis=1))

# === Modellwerte berechnen
psi_mod = psi_model(x_vals, gamma, beta)

# === Fehlermaße
mae = np.mean(np.abs(psi_mod - psi_true))
mse = np.mean((psi_mod - psi_true) ** 2)
corr, _ = pearsonr(psi_mod, psi_true)

# === Ausgabe
print(f"✅ MAE: {mae:.3f}")
print(f"✅ MSE: {mse:.3f}")
print(f"📈 Pearson-Korrelation: r = {corr:.6f}")

# === Plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, psi_true, label="ψ(x) klassisch", color="black", lw=2)
plt.plot(x_vals, psi_mod, label="ψ(x) rekonstruktiv (β)", linestyle="--", color="red")
plt.xlabel("x")
plt.ylabel("ψ(x)")
plt.title("Vergleich: ψ(x) klassisch vs. rekonstruierte Darstellung")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Dateipfade (anpassen, falls nötig) ===
zeta_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
prime_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv"
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"

# === Daten laden (Header überspringen, falls nötig) ===
gamma = pd.read_csv(zeta_path, header=None).iloc[:, 0].astype(float).to_numpy()
primes = pd.read_csv(prime_path, skiprows=1, header=None).iloc[:, 0].astype(int).to_numpy()
beta = pd.read_csv(beta_path, header=None, skiprows=1).iloc[:, 0].astype(float).to_numpy()

# === Echte ψ(x) berechnen (Chebyshev-ψ) ===
psi_true = np.cumsum(np.log(primes))

# === FFF-Modell basierend auf γ und β ===
def psi_reconstruct(x_vals, gamma, beta):
    x_vals = np.array(x_vals)
    rho = 0.5 + 1j * gamma
    out = []
    for x in x_vals:
        val = np.sum(beta * x**rho / (rho * np.log(x + 1e-8)))
        out.append(np.real(val))
    return np.array(out)

# === Vergleichsbereich vorbereiten ===
x_range = primes[:1000]
psi_rec = psi_reconstruct(x_range, gamma[:1000], beta[:1000])
psi_true_range = psi_true[:1000]

# === Abweichung plotten ===
plt.figure(figsize=(10, 5))
plt.plot(x_range, psi_true_range, label="ψ(x) klassisch", lw=2)
plt.plot(x_range, psi_rec, label="ψ(x) rekonstruiert", lw=2, linestyle='--')
plt.xlabel("x")
plt.ylabel("ψ(x)")
plt.title("Vergleich: klassische vs. rekonstruierte ψ(x)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# === Beta-Störungstest ===
np.random.seed(42)
beta_perturbed = beta[:1000] * (1 + 0.01 * np.random.randn(1000))  # 1% Rauschen
psi_perturbed = psi_reconstruct(x_range, gamma[:1000], beta_perturbed)

# === Plot Vergleich
plt.figure(figsize=(10, 4))
plt.plot(x_range, psi_rec - psi_true_range, label="Original β", lw=2)
plt.plot(x_range, psi_perturbed - psi_true_range, label="Gestörtes β", lw=2, linestyle='--')
plt.axhline(0, color='gray', lw=1)
plt.xlabel("x")
plt.ylabel("Abweichung ψ(x) - ψₜᵣᵤₑ")
plt.title("Robustheitstest gegenüber Störung von β(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Operatorgröße ===
N = 300  # Größe des Operators
lambda_mu = 0.2  # Verstärkung der Möbius-Komponente

# === Möbius-Funktion schnell berechnen ===
def mobius_array(n):
    mu = np.ones(n + 1, dtype=int)
    mu[0] = 0
    for i in range(2, n + 1):
        if mu[i] == 1:
            mu[i::i] *= -1
            if i * i <= n:
                mu[i * i::i * i] = 0
    return mu

mu_arr = mobius_array(N)

# === Operator D_mu als Matrix aufbauen ===
Dmu = np.zeros((N, N))
for n in range(2, N + 1):
    arithmetic = lambda_mu * sum(mu_arr[d] * (n // d) for d in range(1, n))
    Dmu[n - 1, n - 1] = arithmetic  # Diagonalelemente

# === Testfunktionen vorbereiten ===
n_vals = np.arange(1, N + 1)
f = np.sin(0.01 * np.pi * n_vals)
g = np.cos(0.01 * np.pi * n_vals)

# === Operatorwirkung berechnen ===
Dmu_f = Dmu @ f
Dmu_g = Dmu @ g

# === Skalarprodukte vergleichen (Selbstadjungiertheitstest) ===
inner1 = np.dot(Dmu_f, g)
inner2 = np.dot(f, Dmu_g)
symmetric = np.allclose(inner1, inner2)

# === Plot zur Visualisierung ===
plt.figure(figsize=(8, 4))
plt.plot(n_vals, Dmu_f, label=r"$D_\mu f$")
plt.plot(n_vals, Dmu_g, label=r"$D_\mu g$")
plt.title("Operatorwirkung auf Testfunktionen")
plt.xlabel("n")
plt.ylabel("Wert")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

print(f"⟨Dμ f, g⟩ = {inner1}")
print(f"⟨f, Dμ g⟩ = {inner2}")
print(f"Selbstadjungiert: {symmetric}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === Euler–Freese-Modellfunktion ===
def beta_series(n, A=1.0, B=0.518, C=0.0):
    """Erweiterte Freese-Folge mit Restglied"""
    return A * n**(-B) + C

# === Summierung bis N ===
def sum_beta(N, A=1.0, B=0.518, C=0.0):
    n_vals = np.arange(1, N + 1)
    beta_vals = beta_series(n_vals, A, B, C)
    return np.sum(beta_vals)

# === Simulation über viele N-Werte hinweg ===
N_max = 50000
N_vals = np.arange(1000, N_max + 1, 1000)
sums = np.array([sum_beta(N) for N in N_vals])
epsilon_vals = 1.0 - sums

# === Plot: Annäherung an ε ===
plt.figure(figsize=(10, 4))
plt.plot(N_vals, epsilon_vals, label='Restwert ε(N)', color='crimson')
plt.axhline(y=0.00012151, color='gray', linestyle='--', label='erwartetes ε ≈ 1/8217')
plt.xlabel("N (Summengrenze)")
plt.ylabel("ε(N) = 1 - ∑ βₙ")
plt.title("Numerische Stabilität der Euler–Freese-Korrektur ε")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Daten einladen ===
# Ersetze durch deinen Pfad zu beta(n)
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta = pd.read_csv(beta_path, names=["drift"], skiprows=1).iloc[:, 0].astype(float).to_numpy()

# === ε(N) berechnen ===
N_vals = np.arange(1000, len(beta), 100)
epsilon_vals = [1 - np.sum(beta[:N]) for N in N_vals]

# === Fit-Modell: ε(N) ~ ε∞ + k / N^p ===
def epsilon_model(N, eps_inf, k, p):
    return eps_inf + k / N**p

initial_guess = [-6, 1, 1]
params, _ = curve_fit(epsilon_model, N_vals, epsilon_vals, p0=initial_guess)
eps_inf, k_fit, p_fit = params

# === Plot ===
plt.figure(figsize=(12, 5))
plt.plot(N_vals, epsilon_vals, label="Berechnetes ε(N)", color="crimson")
plt.plot(N_vals, epsilon_model(N_vals, *params), '--', label=f"Fit: ε∞ ≈ {eps_inf:.5f}, p ≈ {p_fit:.3f}")
plt.axhline(0, color='black', linestyle='dotted')
plt.xlabel("N (Summengrenze)")
plt.ylabel("ε(N) = 1 - Σ βₙ")
plt.title("Fit der Korrektur ε(N) mit Grenzwert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero, diff, zeta, mp
from sympy import mobius
import torch

# Präzision
mp.dps = 40

# Parameter
N = 100  # Dimension des Operators
K = 100  # Anzahl Nullstellen
σ = 0.5  # Breite der Gauß-Approximation
λ = 0.2  # Rückkopplungsfaktor

# Zeta-Nullstellen und Ableitungen
gammas = np.array([float(zetazero(k).imag) for k in range(1, K + 1)])
zeta_derivs = np.array([abs(diff(zeta, 0.5 + 1j * g)) for g in gammas])
weights = 1 / zeta_derivs

# Gaußdelta
def delta_sigma(n, gamma, sigma):
    return np.exp(-((n - gamma) ** 2) / (2 * sigma ** 2))

# Dµ-Matrix erzeugen
D = np.zeros((N, N))

for n in range(1, N + 1):
    for m in range(1, N + 1):
        # Spektraler Anteil
        spectral_sum = sum(weights[k] * delta_sigma(n, gammas[k], σ) if n == m else 0 for k in range(K))

        # Möbius-Rückkopplung
        mobius_sum = sum(mobius(d) * m if d != 0 else 0
                         for d in range(1, n) if n % d == 0 and m == n)

        D[n - 1, m - 1] = spectral_sum + λ * mobius_sum

# Darstellung
plt.imshow(D, cmap='seismic', aspect='auto')
plt.colorbar(label='Wert')
plt.title("Matrixdarstellung von $D_\\mu$")
plt.xlabel("m")
plt.ylabel("n")
plt.tight_layout()
plt.show()

# Optional: Eigenwerte prüfen
eigvals = np.linalg.eigvalsh(D)
plt.plot(eigvals, '.', label='Eigenwerte')
plt.title("Spektrum von $D_\\mu$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# === Bibliotheken ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import mobius
from mpmath import mp, zetazero, diff, zeta
from tqdm import tqdm

# === Präzision ===
mp.dps = 40

# === Parameter ===
N = 1000  # Anzahl Nullstellen
λ = 1.0   # Rückkopplungsparameter
σ = 1.0   # Breite der Gaußfunktion
n_max = 500  # Maximaler Diskretisierungspunkt

# === Zeta-Nullstellen und Gewichte vorberechnen ===
gammas = np.array([float(zetazero(k).imag) for k in range(1, N+1)])
rhos = [0.5 + 1j*g for g in gammas]
zeta_derivs = np.array([abs(diff(zeta, rho)) for rho in rhos])
weights = 1 / zeta_derivs

# === Dμ Operator auf ℓ²(N) definieren ===
def delta_sigma(n, gamma, sigma=σ):
    return np.exp(-((n - gamma)**2) / (2 * sigma**2))

def D_mu(f, n):
    spektral_sum = sum(weights[k] * delta_sigma(n, gammas[k]) * f[n] for k in range(N))
    moebius_sum = sum(mobius(d) * f[n] for d in range(1, n)) if n > 1 else 0
    return spektral_sum + λ * moebius_sum

# === Testvektor: f(n) = 1/n^s ===
s = 0.6
f_vec = np.array([1/n**s if n > 0 else 0 for n in range(n_max)])

# === Wirken von Dμ auf f berechnen ===
D_f = np.zeros(n_max)
for n in tqdm(range(1, n_max)):
    D_f[n] = D_mu(f_vec, n)

# === Plot: Operatorwirkung vs Originalfunktion ===
plt.figure(figsize=(10, 5))
plt.plot(range(1, n_max), f_vec[1:], label=r"$f(n) = 1/n^s$")
plt.plot(range(1, n_max), D_f[1:], label=r"$D_\mu f(n)$", linestyle='--')
plt.xlabel("n")
plt.ylabel("Wert")
plt.title("Operator Dμ auf ℓ²(N)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# === Bibliotheken ===
import numpy as np
import pandas as pd
import torch
from mpmath import mp, diff, zeta
import matplotlib.pyplot as plt

# === Präzision und Gerät ===
mp.dps = 40
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("✅ Device:", device)

# === Parameter ===
N = 1000  # Anzahl Nullstellen (kann später erhöht werden)
sigma = 0.5  # Breite der gaußschen Delta-Näherung
lambda_mu = 1.0  # Rückkopplungsparameter
n_max = 100  # Wirken auf n = 1...n_max

# === Nullstellen aus Datei laden ===
zeros_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
gammas = pd.read_csv(zeros_path, header=None).iloc[:N, 0].astype(float).values
print(f"✅ {len(gammas)} Nullstellen geladen")

# === Ableitungen berechnen (|zeta'(rho)|) ===
rho_vals = [0.5 + 1j * g for g in gammas]
zeta_derivs = [abs(diff(zeta, rho)) for rho in rho_vals]
w_ks = 1 / np.array(zeta_derivs, dtype=np.float64)

# === Testfunktion auf ℓ²(N) ===
f_n = torch.linspace(1, 0, n_max, device=device)

# === Operator Dµ simulieren ===
n_vals = torch.arange(1, n_max+1, dtype=torch.float64, device=device)
Dmu_f = torch.zeros_like(f_n)

for k in range(N):
    gamma_k = gammas[k]
    wk = w_ks[k]
    delta = torch.exp(- (n_vals - gamma_k)**2 / (2 * sigma**2))
    Dmu_f += wk * delta * f_n

# Rückkopplung mit Möbius-Funktion
from sympy import mobius
mu_vec = torch.tensor([mobius(d) for d in range(1, n_max)], dtype=torch.float64, device=device)

for n in range(1, n_max+1):
    sum_mu = sum(mu_vec[d-1] * f_n[n//(d+1)] for d in range(1, n))
    Dmu_f[n-1] += lambda_mu * sum_mu

# === Plot des Operator-Ergebnisses ===
plt.figure(figsize=(10, 4))
plt.plot(n_vals.cpu(), f_n.cpu(), label="Eingabe f(n)")
plt.plot(n_vals.cpu(), Dmu_f.cpu(), label=r"$D_\mu f(n)$", linestyle="--")
plt.xlabel("n")
plt.ylabel("Wert")
plt.title("Wirkung des Operators $D_\\mu$ auf f(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.fft import fft, fftfreq
from mpmath import zeta
import math

# === Skala wählen (z. B. präzise-cleaned) ===
path_beta = "/content/drive/MyDrive/beta_skala_praezision_cleaned.csv"
path_deriv = "/content/drive/MyDrive/zeta_derivatives.csv"

# === FFT-Fensterbreite um jede Primfrequenz ===
epsilon = 0.002

# === Anzahl Primzahlen analysieren ===
num_primes = 20

# === Ableitungen laden ===
df_deriv = pd.read_csv(path_deriv)
zeta_primes = df_deriv["zeta_prime_abs"].astype(float).values
zeta2 = float(zeta(2))
product = zeta2 * zeta_primes

# === Beta-Werte laden ===
df_beta = pd.read_csv(path_beta, header=None)
df_beta = df_beta[pd.to_numeric(df_beta[0], errors='coerce').notna()]
beta_n = df_beta[0].astype(float).values

# === Länge angleichen ===
N = min(len(beta_n), len(product))
beta_n = beta_n[:N]
product = product[:N]

# === Regression & Residuen ===
slope, intercept, *_ = linregress(product, beta_n)
fit = slope * product + intercept
residuals = beta_n - fit

# === FFT der Residuen ===
fft_vals = np.abs(fft(residuals))[:N//2]
freqs = fftfreq(N, d=1)[:N//2]

# === Primzahlen & Frequenzen ===
def prime_log_freqs(num):
    primes, p = [], 2
    while len(primes) < num:
        if all(p % d != 0 for d in range(2, int(p**0.5)+1)):
            primes.append(p)
        p += 1
    return primes, [math.log(p)/(2*math.pi) for p in primes]

primes, prime_freqs = prime_log_freqs(num_primes)

# === Summiere Energie in Umgebung jeder Primfrequenz ===
energies = []
for f in prime_freqs:
    mask = (freqs >= f - epsilon) & (freqs <= f + epsilon)
    energy = np.sum(fft_vals[mask])
    energies.append(energy)

# === Ausgabe als Tabelle ===
df_result = pd.DataFrame({
    "Prime": primes,
    "Freq": [round(f, 5) for f in prime_freqs],
    "Energy": energies
})
display(df_result)

# === Plot: Resonanzbalken ===
plt.figure(figsize=(10, 5))
plt.bar([str(p) for p in primes], energies)
plt.title("Spektrale Resonanzenergie um log(p)/2π")
plt.xlabel("Primzahl p")
plt.ylabel("Summierte FFT-Amplitude")
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
