```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.optimize import minimize
from sklearn.preprocessing import StandardScaler

# === Datenpfade ===
beta_path = '/content/drive/MyDrive/beta_skala_v5_biasfrei_korrigiert.csv'
zeta_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv'
prime_path = '/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv'

# === Daten einlesen ===
beta_df = pd.read_csv(beta_path, header=None, skiprows=1)
beta_raw = pd.to_numeric(beta_df.iloc[:, 1], errors='coerce').dropna().values[:2000]

zeta_raw = pd.read_csv(zeta_path, header=None).squeeze().values[:2000]

prime_df = pd.read_csv(prime_path, header=None, dtype=str)[0]
prime_raw = pd.to_numeric(prime_df, errors='coerce').dropna().values[:2000]
log_primes = np.log(prime_raw)

# === Normierung ===
scaler = StandardScaler()
beta_norm = scaler.fit_transform(beta_raw.reshape(-1, 1)).flatten()
zeta_norm = scaler.fit_transform(zeta_raw.reshape(-1, 1)).flatten()
log_primes_norm = scaler.fit_transform(log_primes.reshape(-1, 1)).flatten()

# === Fourier-Transformation ===
def compute_spectrum(data):
    N = len(data)
    spectrum = np.abs(fft(data))[:N // 2]
    freqs = fftfreq(N)[:N // 2]
    return freqs, spectrum

freqs, beta_spectrum = compute_spectrum(beta_norm)
_, zeta_spectrum = compute_spectrum(zeta_norm)
_, log_primes_spectrum = compute_spectrum(log_primes_norm)

# === Optimierung der Beta-Skala ===
def objective(params):
    scale, shift = params
    adjusted_beta = beta_norm * scale + shift
    _, adjusted_beta_spectrum = compute_spectrum(adjusted_beta)
    loss_zeta = np.sum((adjusted_beta_spectrum - zeta_spectrum) ** 2)
    loss_primes = np.sum((adjusted_beta_spectrum - log_primes_spectrum) ** 2)
    return loss_zeta + loss_primes

initial_params = [1.0, 0.0]
result = minimize(objective, initial_params, method='L-BFGS-B')
optimal_scale, optimal_shift = result.x

# Anwendung der optimalen Transformation
beta_optimized = beta_norm * optimal_scale + optimal_shift
_, beta_optimized_spectrum = compute_spectrum(beta_optimized)

# === Visualisierung ===
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
plt.plot(beta_norm, label='Originale Beta-Skala', color='orange')
plt.plot(beta_optimized, label='Optimierte Beta-Skala', color='red', linestyle='--')
plt.plot(zeta_norm, label='Zeta-Nullstellen', color='green', alpha=0.7)
plt.plot(log_primes_norm, label='log(Primzahlen)', color='blue', alpha=0.7)
plt.title('Zeitbereich: Vergleich der Skalen')
plt.xlabel('Index')
plt.ylabel('Standardisierte Werte')
plt.legend()
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(freqs, beta_spectrum, label='Originale Beta-Skala', color='orange')
plt.plot(freqs, beta_optimized_spectrum, label='Optimierte Beta-Skala', color='red', linestyle='--')
plt.plot(freqs, zeta_spectrum, label='Zeta-Nullstellen', color='green', alpha=0.7)
plt.plot(freqs, log_primes_spectrum, label='log(Primzahlen)', color='blue', alpha=0.7)
plt.title('Frequenzbereich: Fourier-Spektren')
plt.xlabel('Frequenz')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# ========================
# 1. Bibliotheken importieren
# ========================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ========================
# 2. Daten einlesen
# ========================

# Pfade zu den Dateien
beta_path = 'beta_skala_v5_biasfrei_korrigiert.csv'  # CSV mit Spalten: Index, Beta_Skala
zeros_path = 'zeros6.txt'  # TXT oder CSV mit einer Spalte: Zero

# Beta-Skala laden
beta_df = pd.read_csv(beta_path)

# Zeta-Nullstellen laden (z. B. aus LMFDB, Odlyzko oder eigener Liste)
zeros_df = pd.read_csv(zeros_path, header=None, names=['Zero'])

# ========================
# 3. Fehler (Delta) berechnen
# ========================
# Differenz berechnen
beta_df['Delta'] = beta_df['Beta_Skala'] - zeros_df['Zero']

# Gültige Werte mit kleinem Fehler extrahieren (Δ < 1e-4)
valid_df = beta_df[np.abs(beta_df['Delta']) < 1e-4]
valid_indices = valid_df.index.to_numpy()

print(f"Gefundene gültige Werte (Δ < 1e-4): {len(valid_indices)}")

# ========================
# 4. Freese-Funktion definieren
# ========================
# Parameter (aus notarieller Version)
A = 1.882795
beta_exp = 0.916977
C = 2488.144455
B = 1.0
epsilon = 7 / 33300  # Frequenzanteil aus vorheriger Optimierung
omega = 2 * np.pi * epsilon
phi = 0

# Freese-Funktion
def L(N):
    return A * N**beta_exp + C * np.log(N) + B * np.sin(omega * N + phi)

# ========================
# 5. Berechnung der Freese-Reihe
# ========================
N_vals = np.arange(1, len(valid_indices) + 1)
freese_vals = L(N_vals)

# Korrespondierende Zeta-Nullstellen aus zeros_df
zeta_vals = zeros_df.iloc[valid_indices]['Zero'].values

# ========================
# 6. Visualisierung
# ========================
plt.figure(figsize=(10, 6))
plt.plot(N_vals, freese_vals, label='Neue Freese-Reihe', linewidth=2)
plt.plot(N_vals, zeta_vals, '--', label='Zeta-Nullstellen (Referenz)', alpha=0.7)
plt.title('Vergleich: Neue Freese-Reihe vs. Zeta-Nullstellen')
plt.xlabel('Index N')
plt.ylabel('Wert')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ========================
# 7. Statistik der Abweichung
# ========================
abweichung = freese_vals - zeta_vals
print(f"Mittelwert der Abweichung Δ: {np.mean(abweichung):.8f}")
print(f"Standardabweichung: {np.std(abweichung):.8f}")
```

```python
# Schritt 1: Setup
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# CUDA
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Verwendetes Gerät:", device)

# Schritt 2: Zeta-Nullstellen laden
df = pd.read_csv('/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv')
zeta_vals = torch.tensor(df['zero'].values[:2_000_000], dtype=torch.float32, device=device)
n_vals = torch.arange(1, len(zeta_vals)+1, dtype=torch.float32, device=device)

# Schritt 3: Modellklasse definieren
class FreeseModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.A     = nn.Parameter(torch.tensor(1.0))
        self.beta  = nn.Parameter(torch.tensor(0.9))
        self.C     = nn.Parameter(torch.tensor(1.0))
        self.D     = nn.Parameter(torch.tensor(1.0))
        self.gamma = nn.Parameter(torch.tensor(0.5))
        self.omega = nn.Parameter(torch.tensor(0.1))
        self.phi   = nn.Parameter(torch.tensor(0.0))

    def forward(self, n):
        term1 = self.A * torch.pow(n, self.beta)
        term2 = self.C * torch.log(n)
        term3 = self.D * torch.pow(n, -self.gamma) * torch.sin(self.omega * torch.log(n) + self.phi)
        return term1 + term2 + term3

# Schritt 4: Modell, Loss, Optimizer
model = FreeseModel().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.5, patience=500, verbose=True)

# Schritt 5: Training
epochs = 10000
losses = []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(n_vals)
    loss = criterion(outputs, zeta_vals)
    loss.backward()
    optimizer.step()
    scheduler.step(loss)
    losses.append(loss.item())
    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item()}")

# Schritt 6: Ergebnisse
params = {name: param.item() for name, param in model.named_parameters()}
print("\nOptimierte Parameter:")
for k, v in params.items():
    print(f"{k} = {v:.8f}")

# Schritt 7: Plot MSE-Verlauf
plt.figure(figsize=(10,3))
plt.plot(losses)
plt.title("Fehlerverlauf (MSE über Epochen)")
plt.xlabel("Epoche")
plt.ylabel("MSE")
plt.grid(True)
plt.show()

# Schritt 8: Plot Fit vs. Zeta-Nullstellen
with torch.no_grad():
    prediction = model(n_vals).cpu().numpy()
    true_values = zeta_vals.cpu().numpy()

plt.figure(figsize=(12,4))
plt.plot(n_vals.cpu().numpy(), true_values, label="Zeta-Nullstellen")
plt.plot(n_vals.cpu().numpy(), prediction, '--', label="Freese-Modell (GPU)", alpha=0.8)
plt.legend()
plt.title("GPU-Fit: Zeta-Nullstellen vs. Optimiertes Freese-Modell")
plt.xlabel("Index n")
plt.ylabel("Wert")
plt.grid(True)
plt.show()
```

```python
# === Vorbereitung: Bibliotheken laden ===
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# === Gerät auswählen (GPU, falls verfügbar) ===
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Verwendetes Gerät:", device)

# === Daten laden: Zeta-Nullstellen (CSV ohne Header!) ===
df = pd.read_csv('/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv', header=None)
zeta_vals = torch.tensor(df[0].values[:2_000_000], dtype=torch.float32, device=device)
n_vals = torch.arange(1, len(zeta_vals)+1, dtype=torch.float32, device=device)

# === Modellklasse definieren: Erweiterte Freese-Funktion ===
class FreeseModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.A = nn.Parameter(torch.tensor(1.0))
        self.beta = nn.Parameter(torch.tensor(0.9))
        self.C = nn.Parameter(torch.tensor(1.0))
        self.D = nn.Parameter(torch.tensor(1.0))
        self.gamma = nn.Parameter(torch.tensor(0.5))
        self.omega = nn.Parameter(torch.tensor(0.1))
        self.phi = nn.Parameter(torch.tensor(0.0))

    def forward(self, n):
        return (self.A * n ** self.beta
                + self.C * torch.log(n)
                + self.D * n ** (-self.gamma) * torch.sin(self.omega * torch.log(n) + self.phi))

# === Modell initialisieren ===
model = FreeseModel().to(device)

# === Loss-Funktion und Optimierer ===
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# === Training starten ===
epochs = 10_000
loss_history = []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    output = model(n_vals)
    loss = criterion(output, zeta_vals)
    loss.backward()
    optimizer.step()
    loss_history.append(loss.item())

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.6f}")

# === Optimierte Parameter ausgeben ===
params = {name: param.item() for name, param in model.named_parameters()}
print("\nOptimierte Parameter:")
for k, v in params.items():
    print(f"{k:>6} = {v:.8f}")

# === Ergebnisse plotten ===
model.eval()
with torch.no_grad():
    fitted = model(n_vals).cpu().numpy()

plt.figure(figsize=(12, 4))
plt.plot(n_vals.cpu().numpy(), zeta_vals.cpu().numpy(), label='Zeta-Nullstellen')
plt.plot(n_vals.cpu().numpy(), fitted, label='Freese-Modell (GPU)', linestyle='--')
plt.title("GPU-Fit: Zeta-Nullstellen vs. Optimiertes Freese-Modell")
plt.xlabel("Index n")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.show()

# === Fehlerverlauf anzeigen ===
plt.figure(figsize=(12, 4))
plt.plot(loss_history)
plt.title("Fehlerverlauf (MSE über Epochen)")
plt.xlabel("Epoche")
plt.ylabel("MSE")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

# --- Parameter für den Operator ---
N = 1000                      # Diskretisierungspunkte
L = 10                        # Intervall [0, L]
t = np.linspace(1e-5, L, N)   # t-Achse (vermeide t = 0)
dt = t[1] - t[0]              # Schrittweite

# --- Beta-Potenzial definieren (aus Siegel-Theta-Struktur) ---
def beta_potential(t, beta=0.914, B=-0.001, w=-2e-6, phi=9.065):
    return beta + B * np.sin(w * np.log(t) + phi)

V = beta_potential(t)

# --- Kinetischer Term (-i d/dt als Finite-Difference, symmetrisch diskretisiert) ---
kinetic_diag = np.zeros(N)
kinetic_offdiag = -0.5 / dt * np.ones(N-1)

# --- Gesamt-Hamiltonoperator: H = T + V ---
potential_diag = V

# --- Gesamtdiagonale: H = T + V ---
H_diag = kinetic_diag + potential_diag
H_offdiag = kinetic_offdiag

# --- Eigenwerte berechnen (symmetrischer Tridiagonal-Operator) ---
eigvals, eigvecs = eigh_tridiagonal(H_diag, H_offdiag)

# --- Visualisierung ---
plt.figure(figsize=(10, 4))
plt.plot(eigvals[:100], 'o', markersize=3, label='Eigenwerte $H_\\Theta$')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Eigenwertspektrum des Theta-Hamiltonoperators $H_\\Theta$')
plt.xlabel('Index')
plt.ylabel('Eigenwert')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Operation χ – Dirichlet-modulierte Beta-Struktur
import numpy as np
import matplotlib.pyplot as plt
from sympy.ntheory import primitive_root
from sympy import isprime
from math import gcd

# === 1. Dirichlet-Charaktere berechnen (modulo q) ===
def dirichlet_character(n, q, chi_type='principal'):
    if gcd(n, q) != 1:
        return 0  # nicht teilerfremd, dann per Definition 0

    if chi_type == 'principal':
        return 1  # trivialer Charakter

    elif chi_type == 'legendre':
        # (n|q) Legendre-Symbol, nur gültig für ungerade Primzahl q
        return pow(n, (q - 1) // 2, q) if isprime(q) and q % 2 == 1 else 0

    else:
        raise ValueError("Nur 'principal' oder 'legendre' unterstützt.")

# === 2. Beta-Skala: einfaches Oszillationsmodell ===
def beta_scale(n):
    beta_0 = 0.914
    delta = 0.001
    omega = 2e-6
    phi = 9.065
    return beta_0 + delta * np.sin(omega * np.log(n) + phi)

# === 3. Modifiziertes L-Modell ===
def L_chi(s, q, N=10000, chi_type='legendre'):
    result = 0
    for n in range(1, N+1):
        chi_n = dirichlet_character(n, q, chi_type)
        beta_n = beta_scale(n)
        result += chi_n * beta_n / (n**s)
    return result

# === 4. Visualisierung auf kritischer Linie ===
s_vals = [0.5 + 1j * t for t in np.linspace(0.1, 30, 200)]
L_vals = [L_chi(s, q=7, chi_type='legendre') for s in s_vals]

plt.plot([s.imag for s in s_vals], [abs(L) for L in L_vals], label='|Lχ(s)|')
plt.title("Moduliertes Lχ(s) mit Legendre-Charakter χ mod 7")
plt.xlabel("Im(s)")
plt.ylabel("|Lχ(s)|")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Prüfen, ob CUDA verfügbar ist
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Verwendetes Gerät: {device}")

# Parameter
N = 2_000_000  # obere Schranke für die Summe
beta_0 = 0.5
delta = 0.1
omega = 2.0
phi = 0.0

# Dirichlet-Charakter χ mod 7 als Tensor
def dirichlet_chi_mod7(n_tensor):
    # Legendre-Symbol mod 7
    n_mod = n_tensor % 7
    chi = torch.zeros_like(n_tensor, dtype=torch.float32, device=device)
    chi[(n_mod == 1) | (n_mod == 2) | (n_mod == 4)] = 1
    chi[(n_mod == 3) | (n_mod == 5) | (n_mod == 6)] = -1
    return chi

# Werte s = 0.5 + i t für t in [0, 50]
t_vals = torch.linspace(0, 50, 500, device=device)
s_vals = torch.complex(torch.full_like(t_vals, 0.5), t_vals)

# n-Werte vorbereiten
n = torch.arange(1, N + 1, dtype=torch.float32, device=device)

# Beta-Funktion β(n)
def beta_func(n):
    return beta_0 + delta * torch.sin(omega * torch.log(n) + phi)

# Zeta-Freese-Modul mit Charakter χ
def zeta_freese_chi(s_vals):
    chi = dirichlet_chi_mod7(n)
    beta_n = beta_func(n)
    exponent = torch.exp(1j * np.pi * beta_n)

    zeta_results = []

    for s in s_vals:
        term = chi * exponent / n.pow(s)
        zeta_s = torch.sum(term)
        zeta_results.append(zeta_s)

    return torch.stack(zeta_results)

# Berechnung starten
zeta_values = zeta_freese_chi(s_vals)
zeta_abs = torch.abs(zeta_values).cpu().numpy()
zeta_phase = torch.angle(zeta_values).cpu().numpy()
t_vals_cpu = t_vals.cpu().numpy()

# Plotten
plt.figure(figsize=(14, 6))

plt.subplot(2, 1, 1)
plt.plot(t_vals_cpu, zeta_abs, label=r'$|\zeta_{F,\chi}(0.5 + it)|$', color='b')
plt.xlabel(r'$t$')
plt.ylabel(r'$|\zeta_{F,\chi}|$')
plt.title('Betrag der modulierten Zeta-Freese-Funktion (GPU)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t_vals_cpu, zeta_phase, label=r'$\arg(\zeta_{F,\chi})$', color='r')
plt.xlabel(r'$t$')
plt.ylabel(r'$\arg$')
plt.title('Phase der modulierten Zeta-Freese-Funktion (GPU)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

# Gerät wählen: CUDA (GPU), falls verfügbar
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dirichlet-Charakter χ definieren (mod 3, 5, 7)
def dirichlet_chi(n, mod):
    n_mod = n % mod
    if mod == 3:
        return 0 if n_mod == 0 else (1 if n_mod == 1 else -1)
    elif mod == 5:
        return {0: 0, 1: 1, 2: 1j, 3: -1j, 4: -1}[n_mod]
    elif mod == 7:
        return 0 if n_mod == 0 else (1 if pow(n_mod, 3, 7) == 1 else -1)
    return 1

# Theta-Hamiltonoperator für χ aufbauen
def build_theta_operator(N=100, t=1.0, mod=3):
    n = torch.arange(1, N + 1, dtype=torch.float32, device=device).view(-1, 1)
    m = n.T
    chi_vec = torch.tensor([dirichlet_chi(int(k), mod) for k in n.squeeze().tolist()], dtype=torch.cfloat, device=device)
    theta_matrix = chi_vec[:, None] * chi_vec[None, :] * torch.exp(-np.pi * (n**2 + m**2) * t)
    H_theta = -theta_matrix.real
    return H_theta

# Berechne Spektren
operator_data = {}
for mod in [3, 5, 7]:
    H = build_theta_operator(N=100, t=0.2, mod=mod)
    eigvals = torch.linalg.eigvalsh(H).cpu().numpy()
    operator_data[f"χ mod {mod}"] = eigvals

# Visualisierung
plt.figure(figsize=(12, 4))
for label, spectrum in operator_data.items():
    plt.plot(spectrum, '.', label=label)
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)
plt.title("Eigenwerte des Theta-Hamiltonoperators $H_\\Theta^{(\\chi)}$")
plt.xlabel("Index")
plt.ylabel("Eigenwert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# CUDA aktivieren
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dirichlet-Charakter χ mod 7 (Legendre-Symbol)
def legendre_character_mod7(N):
    values = torch.zeros(N, device=device)
    for n in range(1, N + 1):
        if np.gcd(n, 7) != 1:
            values[n - 1] = 0
        elif pow(n, (7 - 1) // 2, 7) == 1:
            values[n - 1] = 1
        else:
            values[n - 1] = -1
    return values

# Modulierter Theta-Hamiltonoperator H_Theta^(chi)
def mod_theta_operator(N, t, chi_values):
    n = torch.arange(1, N + 1, device=device, dtype=torch.float64)
    m = torch.arange(1, N + 1, device=device, dtype=torch.float64)
    n, m = torch.meshgrid(n, m, indexing='ij')
    kernel = torch.exp(-t * (n - m)**2) * chi_values[n.long() - 1] * chi_values[m.long() - 1]
    return kernel

# Setup
N = 100
t_values = torch.linspace(0.01, 1.5, 100, device=device)
chi = legendre_character_mod7(N)

# Plot
fig, ax = plt.subplots()
line, = ax.plot([], [], 'bo', markersize=3)
ax.set_xlim(0, N)
ax.set_ylim(-0.5, 0.1)
ax.set_title("Dynamik der Eigenwerte von $H_\\Theta^{(\\chi)}(t)$")
ax.set_xlabel("Index")
ax.set_ylabel("Eigenwert")

def update(frame):
    t = t_values[frame].item()
    H = mod_theta_operator(N, t, chi)
    eigvals = torch.linalg.eigvalsh(H).real
    line.set_data(np.arange(N), eigvals.cpu().numpy())
    ax.set_title(f"$H_\\Theta^{{(\\chi)}}(t)$ bei t = {t:.2f}")
    return line,

ani = FuncAnimation(fig, update, frames=len(t_values), blit=True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
from scipy.fft import fft, fftfreq

# Einstellungen
N = 100
t_indices = np.linspace(0.01, 0.2, 10)
start_index = 70

# Dirichlet-Charakter χ mod 7
def dirichlet_chi_mod7(n):
    n = n % 7
    return {
        1: 1,
        2: 1,
        3: -1,
        4: 1,
        5: -1,
        6: -1
    }.get(n, 0)

# Hamilton-Matrix H_Θ^χ(t)
def theta_hamiltonian(N, t):
    H = np.zeros((N, N))
    for n in range(1, N+1):
        for m in range(1, N+1):
            chi_n = dirichlet_chi_mod7(n)
            chi_m = dirichlet_chi_mod7(m)
            if chi_n == 0 or chi_m == 0:
                continue
            value = chi_n * chi_m * np.exp(-((n - m)**2) * t)
            H[n-1, m-1] = value
    return H

# Spektralanalyse für t-Werte
spectra = []
for t_val in t_indices:
    H = theta_hamiltonian(N, t_val)
    eigvals = np.linalg.eigvalsh(H)
    spectra.append(eigvals[start_index:])

spectra = np.array(spectra).T

# Fourier-Analyse über t
freqs = fftfreq(len(t_indices), d=(t_indices[1] - t_indices[0]))
fft_data = np.abs(fft(spectra, axis=1))

# Plot
fig, axs = plt.subplots(2, 1, figsize=(12, 6), sharex=True)

im = axs[0].imshow(spectra, aspect='auto', extent=[t_indices[0], t_indices[-1], start_index, N], origin='lower')
axs[0].set_title('Eigenwerte (Index ≥ 70) über t')
axs[0].set_ylabel('Index')
fig.colorbar(im, ax=axs[0], label='Eigenwert')

im2 = axs[1].imshow(fft_data[:, 1:20], aspect='auto', extent=[freqs[1], freqs[19], start_index, N], origin='lower', cmap='viridis')
axs[1].set_title('Fourier-Spektrum der Eigenwerte')
axs[1].set_xlabel('Frequenz')
axs[1].set_ylabel('Index')
fig.colorbar(im2, ax=axs[1], label='Amplitude')

plt.tight_layout()
plt.show()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# --- Parameter ---
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Verwendetes Gerät:", device)

N = 100  # Matrixgröße
start_index = 70  # Betrachteter Indexbereich (z.B. Index ≥ 70)
t_indices = np.linspace(0.01, 0.2, 50)  # Feiner aufgelöst

# Dirichlet-Charakter χ mod 7 (Legendre-Symbol)
modulus = 7
chi_values = [pow(n, (modulus - 1) // 2, modulus) if np.gcd(n, modulus) == 1 else 0 for n in range(modulus)]
chi_values = [1 if x == 1 else -1 if x == modulus - 1 else 0 for x in chi_values]

def chi(n):
    return chi_values[n % modulus]

# Theta-Hamiltonoperator als Matrix
def theta_operator_matrix(t):
    H = torch.zeros((N, N), dtype=torch.cfloat, device=device)
    for n in range(1, N + 1):
        for m in range(1, N + 1):
            arg = -np.pi * ((n - m) ** 2) / t
            H[n - 1, m - 1] = chi(n) * chi(m) * torch.exp(torch.tensor(arg, dtype=torch.cfloat))
    return H

# --- Hauptanalyse ---
spectra = []
for t in t_indices:
    H = theta_operator_matrix(t)
    eigvals = torch.linalg.eigvals(H).real.cpu().numpy()
    spectra.append(eigvals[start_index:])  # Nur hohe Indizes
spectra = np.array(spectra).T  # shape: [indices, t]

# Fourier-Spektrum entlang t-Achse
fft_data = np.abs(fft(spectra, axis=1))
freqs = fftfreq(len(t_indices), d=(t_indices[1] - t_indices[0]))

# Nur positive Frequenzen (z. B. bis Index 10)
max_freq_index = min(10, fft_data.shape[1] - 1)

# --- Visualisierung ---
fig, axs = plt.subplots(2, 1, figsize=(14, 6), sharex=True)

# Spektrum: Eigenwertverlauf
im = axs[0].imshow(spectra, aspect='auto', extent=[t_indices[0], t_indices[-1], start_index, N], origin='lower')
axs[0].set_title('Eigenwerte (Index ≥ {}) über t'.format(start_index))
axs[0].set_ylabel('Index')
fig.colorbar(im, ax=axs[0], label='Eigenwert')

# Fourier-Spektrum
im2 = axs[1].imshow(fft_data[:, 1:max_freq_index+1], aspect='auto',
                    extent=[freqs[1], freqs[max_freq_index], start_index, N], origin='lower', cmap='viridis')
axs[1].set_title('Fourier-Spektrum der Eigenwerte')
axs[1].set_xlabel('Frequenz')
axs[1].set_ylabel('Index')
fig.colorbar(im2, ax=axs[1], label='Amplitude')

plt.tight_layout()
plt.show()
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
N = 100
t_vals = torch.linspace(0.01, 2, 200, device=device)

def theta_operator_matrix(t, mod=7):
    n = torch.arange(1, N+1, device=device).float()
    chi = torch.tensor([(1 if pow(k, (mod-1)//2, mod) == 1 else -1) if np.gcd(k, mod) == 1 else 0 for k in range(1, N+1)], device=device).float()
    diag = chi * torch.exp(-t * n**2)
    return torch.diag(diag)

# Eigenwertmatrix über t
eigenvalues = []
for t in t_vals:
    H = theta_operator_matrix(t)
    evals = torch.linalg.eigvalsh(H)
    eigenvalues.append(evals.cpu().numpy())

eigenvalues = np.array(eigenvalues).T  # [Index, Zeit]
```

```python
import pywt

index_range = range(75, 100)
wavelet_result = []

for i in index_range:
    coeffs, freqs = pywt.cwt(eigenvalues[i], scales=np.arange(1, 128), wavelet='cmor')
    wavelet_result.append(np.abs(coeffs))

wavelet_result = np.array(wavelet_result)
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 6))

im = axs.imshow(
    wavelet_result[:, :, :150].mean(axis=0),
    extent=[t_vals[0].item(), t_vals[-1].item(), 1, 128],
    aspect='auto',
    origin='lower',
    cmap='plasma'
)

axs.set_title('Wavelet-Zeit-Frequenz-Darstellung (Index 75–99)')
axs.set_xlabel('t')
axs.set_ylabel('Skalierung (Frequenz ~ 1/Skala)')
fig.colorbar(im, ax=axs, label='Amplitude')
plt.tight_layout()
plt.show()
```
