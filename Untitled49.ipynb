```python
# üìå Notwendige Bibliotheken importieren
import torch
import requests
import matplotlib.pyplot as plt

# üìå Ger√§t ausw√§hlen (GPU bevorzugt, falls verf√ºgbar)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"üîç Nutze Ger√§t: {device}")

# üìå Zeta-Nullstellen von Odlyzko's Datenbank laden
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(URL)
zeta_zeros = torch.tensor(
    [float(line.strip()) for line in response.text.split("\n") if line.strip()],
    device=device
)

# üìå Anzahl der Nullstellen w√§hlen (N=100000, f√ºr Spurformel-Test mit 500 Eigenwerten)
N = min(100000, len(zeta_zeros))
zeta_zeros = zeta_zeros[:N]
print(f"‚úÖ Erfolgreich {N} Zeta-Nullstellen geladen!")

# üìå Sparse Hamilton-Operator definieren
diagonal = zeta_zeros
off_diagonal = torch.ones(N-1, device=device)
H = torch.diag(diagonal) + torch.diag(off_diagonal, 1) + torch.diag(off_diagonal, -1)

# üìå Eigenwerte mit PyTorch berechnen (Batchweise)
batch_size = 5000
eigenvalues = []
for i in range(0, N, batch_size):
    batch_H = H[i:i+batch_size, i:i+batch_size]
    batch_eigenvalues = torch.linalg.eigvalsh(batch_H)
    eigenvalues.append(batch_eigenvalues)

eigenvalues = torch.cat(eigenvalues)[:500]  # Nur die ersten 500 Eigenwerte f√ºr Beta-Test
print(f"‚úÖ Erste 10 Eigenwerte: {eigenvalues[:10].cpu().numpy()}")

# üìå Spurformel berechnen f√ºr verschiedene t-Werte
t_values = torch.linspace(0.01, 10, 100, device=device)
trace_values = torch.tensor(
    [torch.sum(torch.exp(-t * eigenvalues)) for t in t_values], device=device
)

# üìå Beta-Korrektur anwenden
beta = 1/137  # Feinstrukturkonstante als Beta-Korrektur
corrected_trace = trace_values * torch.exp(beta * torch.pi * 1j)

# üìå Fehleranalyse: Abweichung zwischen Original und Korrektur
error = torch.abs(trace_values - corrected_trace)

# üìå Plots
plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), trace_values.cpu().real, 'b-', label="Numerische Spurformel")
plt.plot(t_values.cpu(), corrected_trace.cpu().real, 'r--', label=r"Mit $e^{\beta \pi i}$-Korrektur")
plt.xlabel("t-Wert")
plt.ylabel("Spurformel-Summe")
plt.title("Vergleich der Spurformel mit und ohne Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), error.cpu(), 'k-', label="Abweichung")
plt.axhline(0, color='r', linestyle='dashed', label="Perfekte √úbereinstimmung")
plt.xlabel("t-Wert")
plt.ylabel("Fehler")
plt.title("Fehleranalyse der Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()
```

```python
# üìå Notwendige Bibliotheken importieren
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import requests
import matplotlib.pyplot as plt

# üìå Zeta-Nullstellen von Odlyzko's Datenbank laden
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(URL)
zeta_zeros = np.array([float(line.strip()) for line in response.text.split("\n") if line.strip()])

# üìå Anzahl der Nullstellen w√§hlen (maximal 100000)
N = min(100000, len(zeta_zeros))
zeta_zeros = zeta_zeros[:N]
print(f"‚úÖ Erfolgreich {N} Zeta-Nullstellen geladen!")

# üìå Sparse Hamilton-Operator definieren (NUR CPU!)
diag_values = zeta_zeros
off_diag_values = np.ones(N-1)

H_sparse = sp.diags(
    diagonals=[diag_values, off_diag_values, off_diag_values],
    offsets=[0, -1, 1],
    format="csc"
)

# üìå Eigenwerte berechnen mit Sparse-Matrix (scipy nutzt **CPU**)
print(f"‚è≥ Berechne {N} Eigenwerte...")
eigenvalues, _ = spla.eigsh(H_sparse, k=500, which="SM")  # **Nur 500 kleinste Eigenwerte**
print(f"‚úÖ Erste 10 Eigenwerte: {eigenvalues[:10]}")

# üìå Spurformel berechnen f√ºr verschiedene t-Werte
t_values = np.linspace(0.01, 10, 100)
trace_values = np.array([np.sum(np.exp(-t * eigenvalues)) for t in t_values])

# üìå Beta-Korrektur anwenden
beta = 1/137  # Feinstrukturkonstante als Beta-Korrektur
corrected_trace = trace_values * np.exp(beta * np.pi * 1j)

# üìå Fehleranalyse: Abweichung zwischen Original und Korrektur
error = np.abs(trace_values - corrected_trace.real)

print(f"‚úÖ Maximale Abweichung: {max_error:.10f}")
print(f"‚úÖ Durchschnittliche Abweichung: {mean_error:.10f}")

# Falls error ein NumPy-Array ist, konvertiere es zu einem Tensor
error_tensor = torch.tensor(error, device=device)

# üìå Numerische Analyse der Abweichung
max_error = torch.max(error_tensor).item()
mean_error = torch.mean(error_tensor).item()

print(f"‚úÖ Maximale Abweichung: {max_error:.10f}")
print(f"‚úÖ Durchschnittliche Abweichung: {mean_error:.10f}")

# üìå Plots
plt.figure(figsize=(8,5))
plt.plot(t_values, trace_values.real, 'b-', label="Numerische Spurformel")
plt.plot(t_values, corrected_trace.real, 'r--', label=r"Mit $e^{\beta \pi i}$-Korrektur")
plt.xlabel("t-Wert")
plt.ylabel("Spurformel-Summe")
plt.title("Vergleich der Spurformel mit und ohne Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t_values, error, 'k-', label="Abweichung")
plt.axhline(0, color='r', linestyle='dashed', label="Perfekte √úbereinstimmung")
plt.xlabel("t-Wert")
plt.ylabel("Fehler")
plt.title("Fehleranalyse der Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()
```

```python
import torch
import scipy.sparse as sp
import requests
import matplotlib.pyplot as plt

# üìå Ger√§t w√§hlen
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"üîç Nutze Ger√§t: {device}")

# üìå Zeta-Nullstellen von Odlyzko's Datenbank laden
URL = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
response = requests.get(URL)
zeta_zeros = torch.tensor([float(line.strip()) for line in response.text.split("\n") if line.strip()], device=device)

# üìå Anzahl der Nullstellen w√§hlen
N = min(100000, len(zeta_zeros))
zeta_zeros = zeta_zeros[:N]
print(f"‚úÖ Erfolgreich {N} Zeta-Nullstellen geladen!")

# üìå Sparse Hamilton-Operator definieren
diagonal = zeta_zeros
off_diagonal = torch.ones(N-1, device=device)
H = torch.diag(diagonal) + torch.diag(off_diagonal, 1) + torch.diag(off_diagonal, -1)

# üìå Eigenwerte mit PyTorch berechnen (Optimierung: Direkt in Tensor schreiben)
batch_size = 5000
eigenvalues = torch.empty(N, device=device)  # Speicher reservieren

for i in range(0, N, batch_size):
    batch_H = H[i:i+batch_size, i:i+batch_size]
    batch_eigenvalues = torch.linalg.eigvalsh(batch_H)
    eigenvalues[i:i+batch_size] = batch_eigenvalues

print(f"‚úÖ Erste 10 Eigenwerte: {eigenvalues[:10].cpu().numpy()}")

# üìå Spurformel effizient berechnen (Performance-Boost mit `einsum`)
t_values = torch.linspace(0.01, 10, 100, device=device)
trace_values = torch.einsum("i,i->", torch.exp(-t_values[:, None] * eigenvalues), torch.ones_like(eigenvalues))

# üìå Beta-Korrektur anwenden
beta = 1/137
corrected_trace = trace_values * torch.exp(beta * torch.pi * 1j)

# üìå Fehleranalyse
error = torch.abs(trace_values - corrected_trace)

# üìå Plots f√ºr Reelle & Imagin√§re Spurformel
plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), trace_values.cpu().real, 'b-', label="Reell: Spurformel")
plt.plot(t_values.cpu(), corrected_trace.cpu().real, 'r--', label=r"Reell: $e^{\beta \pi i}$-Korrektur")
plt.xlabel("t-Wert")
plt.ylabel("Reeller Teil der Spur")
plt.title("Vergleich der Spurformel mit Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), trace_values.cpu().imag, 'b-', label="Imagin√§r: Spurformel")
plt.plot(t_values.cpu(), corrected_trace.cpu().imag, 'r--', label=r"Imagin√§r: $e^{\beta \pi i}$-Korrektur")
plt.xlabel("t-Wert")
plt.ylabel("Imagin√§rer Teil der Spur")
plt.title("Vergleich der Imagin√§rteile")
plt.legend()
plt.grid()
plt.show()

# üìå Fehlerplot
plt.figure(figsize=(8,5))
plt.plot(t_values.cpu(), error.cpu(), 'k-', label="Abweichung")
plt.axhline(0, color='r', linestyle='dashed', label="Perfekte √úbereinstimmung")
plt.xlabel("t-Wert")
plt.ylabel("Fehler")
plt.title("Fehleranalyse der Beta-Korrektur")
plt.legend()
plt.grid()
plt.show()
```
