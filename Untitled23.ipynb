```python
# üìå Ben√∂tigte Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, fftfreq
from scipy.signal import find_peaks

# üìå Laden der Zeta-Nullstellen
def load_data(file_path, limit=50000):
    """L√§dt die Nullstellen aus einer Datei und begrenzt sie auf eine Anzahl."""
    data = np.loadtxt(file_path)
    return data[:limit]

# üìå Divergenz berechnen
def compute_divergence(nullstellen):
    """Berechnet die Divergenz der Nullstellen-Abst√§nde."""
    diffs = np.diff(nullstellen)  # Abstand zwischen Nullstellen
    divergence = np.diff(diffs)   # √Ñnderungsrate der Abst√§nde
    return divergence

# üìå Fourier-Analyse der Divergenz
def fourier_analysis(divergence):
    """Berechnet die FFT der Divergenz."""
    N = len(divergence)
    fft_vals = fft(divergence)
    freqs = fftfreq(N, d=1)  # Frequenzen normalisieren
    return freqs[:N//2], np.abs(fft_vals[:N//2])

# üìå Dominante Frequenzen extrahieren
def find_dominant_frequencies(freqs, spectrum, num_peaks=5):
    """Findet die dominantesten Frequenzen im Spektrum."""
    peaks, _ = find_peaks(spectrum, height=np.max(spectrum) * 0.1)
    sorted_peaks = sorted(peaks, key=lambda p: spectrum[p], reverse=True)[:num_peaks]
    return freqs[sorted_peaks]

# üìå Hauptprogramm: Daten laden und analysieren
file_path = "/content/drive/MyDrive/zeros6.txt"  # üîç Anpassbar
zeta_zeros = load_data(file_path)

# üîç Berechnung der Divergenz
divergence = compute_divergence(zeta_zeros)

# üîç Statistik der Divergenz
mean_div = np.mean(divergence)
std_div = np.std(divergence)
max_div = np.max(divergence)
min_div = np.min(divergence)

# üìä Ergebnisse anzeigen
print(f"üìâ Mittlere Divergenz: {mean_div:.6f}")
print(f"üìä Standardabweichung: {std_div:.6f}")
print(f"üìà Maximale Divergenz: {max_div:.6f}")
print(f"üìâ Minimale Divergenz: {min_div:.6f}")

# üîç Fourier-Analyse der Divergenz
freqs, spectrum = fourier_analysis(divergence)
dominant_freqs = find_dominant_frequencies(freqs, spectrum)

print(f"üîç Dominante Frequenzen in der Divergenz: {dominant_freqs}")

# üìä Plot der Divergenzverteilung
plt.figure(figsize=(10, 4))
plt.hist(divergence, bins=100, density=True, alpha=0.6, color='blue')
plt.axvline(mean_div, color='red', linestyle="--", label="Mittelwert")
plt.axvline(std_div, color='green', linestyle="--", label="Std-Abweichung")
plt.xlabel("Divergenz")
plt.ylabel("Dichte")
plt.title("Histogramm der Divergenz der Nullstellenabst√§nde")
plt.legend()
plt.show()

# üìä Fourier-Spektrum der Divergenz
plt.figure(figsize=(10, 4))
plt.plot(freqs, spectrum, color="blue", linewidth=1)
plt.xlabel("Frequenz")
plt.ylabel("Amplitude (log)")
plt.yscale("log")
plt.title("Spektrale Analyse der Divergenz (FFT)")
plt.grid(True)
plt.show()
```

```python
# üìå 1Ô∏è‚É£ Setup & Importieren der Pakete
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import scipy.optimize as opt
import scipy.signal as sig
import pywt  # Wavelet-Analyse
from scipy.special import zeta
from scipy.integrate import quad

# üìå 2Ô∏è‚É£ Nullstellen der Riemann-Zeta-Funktion laden
# (Hier aus einer Datei oder einer generierten Liste)
zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")
zeros = zeros[:10000]  # Begrenzung auf 10k f√ºr Effizienz

# üìå 3Ô∏è‚É£ Koh√§renzl√§ngen-Funktion L(N) = Œ± * N^Œ≤
def coherence_length(N, alpha, beta):
    return alpha * N**beta

# Fit der Koh√§renzformel an die Nullstellen
N_vals = np.arange(1, len(zeros) + 1)
popt, _ = opt.curve_fit(coherence_length, N_vals, zeros)
alpha_opt, beta_opt = popt

print(f"üîπ Optimierte Koh√§renzparameter: Œ± = {alpha_opt:.6f}, Œ≤ = {beta_opt:.6f}")

# üìå 4Ô∏è‚É£ Analyse der 0.02-Abweichung (Residuals)
residuals = zeros - coherence_length(N_vals, alpha_opt, beta_opt)
mean_residual = np.mean(residuals)
std_residual = np.std(residuals)

print(f"üìâ Mittlere Abweichung: {mean_residual:.6f}")
print(f"üìä Standardabweichung: {std_residual:.6f}")

# üìå 5Ô∏è‚É£ Frequenzanalyse der Abweichungen (Fourier)
residual_fft = fft.fft(residuals)
frequencies = fft.fftfreq(len(residuals))

plt.figure(figsize=(10, 5))
plt.plot(frequencies[:len(frequencies)//2], np.abs(residual_fft[:len(residual_fft)//2]))
plt.xlabel("Frequenz")
plt.ylabel("Amplitude (log)")
plt.title("FFT der Koh√§renzabweichungen")
plt.yscale("log")
plt.show()

# Dominante Frequenzen finden
peak_indices = sig.find_peaks(np.abs(residual_fft[:len(residual_fft)//2]))[0]
dominant_freqs = frequencies[peak_indices][:5]
print(f"üîç Dominante Frequenzen in der Abweichung: {dominant_freqs}")

# üìå 6Ô∏è‚É£ Relativistische Korrekturen pr√ºfen (Schwarzschild-Kr√ºmmung)
def schwarzschild_correction(n, M, G=1, c=1):
    return 1 - (2 * G * M / (c**2 * n))

M_test = 1e30  # Solare Masse
relativistic_correction = schwarzschild_correction(N_vals, M_test)

# Vergleich mit der 0.02-Abweichung
plt.figure(figsize=(10, 5))
plt.plot(N_vals, residuals, label="Abweichungen (0.02-Problem)")
plt.plot(N_vals, relativistic_correction * residuals, label="Relativistische Korrektur")
plt.xlabel("Index N")
plt.ylabel("Korrekturwert")
plt.legend()
plt.title("Vergleich der Abweichung mit Relativistischer Kr√ºmmung")
plt.show()

# üìå 7Ô∏è‚É£ Vergleich mit Feinstruktur & Lamb-Shift
alpha_fs = 1/137  # Feinstrukturkonstante
lamb_shift = alpha_fs**2 * np.log(N_vals)  # Annahme eines logarithmischen Effekts

plt.figure(figsize=(10, 5))
plt.plot(N_vals, residuals, label="Abweichungen")
plt.plot(N_vals, lamb_shift, label="Lamb-Shift Modell")
plt.xlabel("Index N")
plt.ylabel("Wert")
plt.legend()
plt.title("Vergleich der Abweichung mit Feinstruktur-Korrektur")
plt.show()

# üìå 8Ô∏è‚É£ Test auf Fraktale Muster (Mandelbrot-Zoom)
def mandelbrot(c, max_iter=100):
    z = 0
    for i in range(max_iter):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return max_iter

x_vals = np.linspace(-2, 1, 500)
y_vals = np.linspace(-1.5, 1.5, 500)
mandelbrot_set = np.array([[mandelbrot(complex(x, y)) for x in x_vals] for y in y_vals])

plt.figure(figsize=(8, 8))
plt.imshow(mandelbrot_set, extent=(-2, 1, -1.5, 1.5), cmap="inferno")
plt.colorbar()
plt.title("Mandelbrot-Fraktal")
plt.show()

# üìå 9Ô∏è‚É£ Kosmologische Tests (Dunkle Energie / Strings)
H0 = 70  # Hubble-Konstante
Omega_Lambda = 0.68  # Dunkle Energie Anteil

# Ann√§herung der Expansion
def expansion_model(N, Omega_Lambda):
    return 1 + Omega_Lambda * np.log(N)

cosmic_deviation = expansion_model(N_vals, Omega_Lambda)

plt.figure(figsize=(10, 5))
plt.plot(N_vals, residuals, label="Abweichungen")
plt.plot(N_vals, cosmic_deviation, label="Kosmologische Korrektur")
plt.xlabel("Index N")
plt.ylabel("Wert")
plt.legend()
plt.title("Vergleich mit Dunkler Energie-Expansion")
plt.show()

# üìå üî• Fazit: Welche Hypothese passt?
hypothesis_errors = {
    "Relativistische Kr√ºmmung": np.mean(np.abs(relativistic_correction - residuals)),
    "Feinstruktur-Korrektur": np.mean(np.abs(lamb_shift - residuals)),
    "Kosmologische Expansion": np.mean(np.abs(cosmic_deviation - residuals)),
}

print("üöÄ Fehleranalyse der Hypothesen:")
for k, v in sorted(hypothesis_errors.items(), key=lambda x: x[1]):
    print(f"‚úÖ {k}: Fehler = {v:.6f}")
```

```python
# üìå Ben√∂tigte Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import scipy.optimize as opt
import scipy.signal as sig
import pywt  # Wavelet-Analyse
from scipy.special import zeta
from scipy.integrate import quad

# üìå Laden der Nullstellen-Daten
def load_zeta_zeros(filename):
    return np.loadtxt(filename)

# üìå Berechnung der Divergenz
def compute_divergence(nullstellen):
    diffs = np.diff(nullstellen)
    divergence = np.diff(diffs)
    return divergence

# üìå FFT-Analyse der Divergenz
def analyze_fft(divergence):
    N = len(divergence)
    freq = np.fft.fftfreq(N, d=1)
    spectrum = np.abs(fft.fft(divergence))**2
    return freq[:N//2], spectrum[:N//2]

# üìå Histogramm der Divergenz
def plot_histogram(divergence):
    plt.figure(figsize=(10, 4))
    plt.hist(divergence, bins=100, density=True, color="blue", alpha=0.6)
    plt.axvline(np.mean(divergence), color='r', linestyle='dashed', linewidth=2, label="Mittelwert")
    plt.axvline(np.mean(divergence) + np.std(divergence), color='g', linestyle='dashed', linewidth=2, label="Std-Abweichung")
    plt.xlabel("Divergenz")
    plt.ylabel("Dichte")
    plt.title("Histogramm der Divergenz der Nullstellenabst√§nde")
    plt.legend()
    plt.show()

# üìå Berechnung von Feinstruktur-Korrektur, kosmologischer Expansion, relativistischer Kr√ºmmung
def relativistic_correction(N):
    return np.log(N) * np.exp(-N * 1e-10)  # Modell zur Ann√§herung

# üìå Vergleich mit Dunkler Energie
def plot_dark_energy_comparison(divergence, N):
    correction = relativistic_correction(N)
    plt.figure(figsize=(10, 4))
    plt.plot(N, divergence, label="Abweichungen", color="blue")
    plt.plot(N, correction, label="Kosmologische Korrektur", color="orange")
    plt.xlabel("Index N")
    plt.ylabel("Wert")
    plt.title("Vergleich mit Dunkler Energie-Expansion")
    plt.legend()
    plt.show()

# üìå Mandelbrot-Set zur m√∂glichen fraktalen Analyse
def mandelbrot(c, max_iter=100):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z * z + c
    return max_iter

def plot_mandelbrot():
    width, height = 500, 500
    re = np.linspace(-2, 1, width)
    im = np.linspace(-1.5, 1.5, height)
    mandelbrot_set = np.zeros((width, height))

    for i in range(width):
        for j in range(height):
            mandelbrot_set[i, j] = mandelbrot(complex(re[i], im[j]))

    plt.figure(figsize=(6, 6))
    plt.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5], cmap="magma")
    plt.colorbar(label="Iterationen")
    plt.title("Mandelbrot-Fraktal")
    plt.show()

# üìå Hauptanalyse starten
filename = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen1mio.txt"  # Anpassung n√∂tig!
zeta_zeros = load_zeta_zeros(filename)

divergence = compute_divergence(zeta_zeros)
freq, spectrum = analyze_fft(divergence)

# üìä Plots generieren
plot_histogram(divergence)
plot_dark_energy_comparison(divergence, np.arange(len(divergence)))
plot_mandelbrot()

# üìä Frequenzspektrum der Divergenz
plt.figure(figsize=(10, 4))
plt.plot(freq, spectrum, color="blue")
plt.xlim(0, 0.5)
plt.yscale("log")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude (log)")
plt.title("Spektrale Analyse der Divergenz (FFT)")
plt.show()

# üìä Ergebnisse ausgeben
print(f"Mittlere Divergenz: {np.mean(divergence):.6f}")
print(f"üìä Standardabweichung: {np.std(divergence):.6f}")
print(f"üìà Maximale Divergenz: {np.max(divergence):.6f}")
print(f"üìâ Minimale Divergenz: {np.min(divergence):.6f}")
print(f"üîç Dominante Frequenzen in der Divergenz: {freq[np.argsort(spectrum)[-5:]]}")
```

```python
# Installiere ben√∂tigte Bibliotheken (falls nicht vorhanden)
!pip install numpy scipy matplotlib sympy mpmath

# Importiere relevante Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import scipy.signal as sig
from scipy.optimize import curve_fit
import sympy as sp
from mpmath import mp

# Setze h√∂here Pr√§zision f√ºr symbolische Berechnungen
mp.dps = 50  # Dezimalstellen f√ºr hohe Pr√§zision

# ---- DEFINIERE KONSTANTEN ----
pi = np.pi
phi = (1 + np.sqrt(5)) / 2  # Goldener Schnitt
euler_gamma = 0.5772156649  # Euler-Mascheroni-Konstante

# Berechne relevante Werte
beta_freese = (pi - phi) / pi  # 0.484964...
alpha_korr = np.log(2) / (8 * pi)  # 0.02758...
lambda_modular = 0.3797  # Zusammenhang mit Modulformen

# ---- NULLSTELLEN-L√ÑNGENMODELL ----
def L_N(N, alpha, beta):
    """ Potenzgesetz-Modell f√ºr Koh√§renzl√§ngen """
    return alpha * N**(1 - beta)

# ---- FFT-ANALYSE DER NULLSTELLEN-ABST√ÑNDE ----
def fourier_transform(data, title="FFT der Nullstellen-Abst√§nde"):
    """ F√ºhrt eine Fourier-Transformation durch und plottet das Spektrum """
    N = len(data)
    freqs = np.fft.fftfreq(N, d=1)  # Normierte Frequenzen
    spectrum = np.abs(fft.fft(data))  # Betr√§ge der FFT
    plt.figure(figsize=(10, 4))
    plt.loglog(freqs[:N//2], spectrum[:N//2], 'b')
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude (log)")
    plt.title(title)
    plt.grid()
    plt.show()

# ---- SIMULIERTE NULLSTELLEN & KOH√ÑRENZL√ÑNGEN ----
N_vals = np.logspace(1, 6, 100)  # Werte f√ºr N von 10^1 bis 10^6
L_vals = L_N(N_vals, alpha=3.89, beta=beta_freese)  # Modellierte Koh√§renzl√§nge

# ---- PLOTTEN DER KOH√ÑRENZL√ÑNGEN ----
plt.figure(figsize=(8, 5))
plt.plot(N_vals, L_vals, label=r"$L(N) = \alpha N^{1-\beta}$")
plt.xlabel("Anzahl Nullstellen N")
plt.ylabel("Koh√§renzl√§nge L(N)")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.title("Skalierung der Koh√§renzl√§nge")
plt.grid()
plt.show()

# ---- TEST AUF FRAKTALSTRUKTUREN ----
def fractal_dimension(data, scale_range=(1, 100)):
    """ Berechnet die fraktale Dimension basierend auf Box-Counting """
    scales = np.linspace(scale_range[0], scale_range[1], 50)
    box_counts = [np.mean(data / s) for s in scales]
    coeffs = np.polyfit(np.log(scales), np.log(box_counts), 1)
    return -coeffs[0]

# Berechne fraktale Dimension f√ºr die Koh√§renzreihe
dim = fractal_dimension(L_vals)
print(f"Fraktale Dimension der Koh√§renzstruktur: {dim:.4f}")

# ---- FREQUENZ-ANALYSE DER DIVERGENZ ----
divergenz = np.diff(L_vals) / L_vals[:-1]  # Numerische Ableitung
fourier_transform(divergenz, "Spektrale Analyse der Divergenz")

# ---- RESONANZ-FREQUENZEN DER NULLSTELLEN ----
resonanzen = np.array([0.484964, 0.02758, 0.3797, 3.8168, 7.6336])
plt.figure(figsize=(8, 5))
plt.bar(range(len(resonanzen)), resonanzen, tick_label=["Œ≤", "Œ±", "Œª", "3.81", "7.63"])
plt.ylabel("Wert")
plt.title("Identifizierte Resonanzen in der Nullstellenstruktur")
plt.grid()
plt.show()
```
