```python
# üìå Wichtige Bibliotheken installieren
!pip install numpy scipy matplotlib sympy tqdm

# üìö Importieren der Bibliotheken
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
import sympy as sp
from tqdm import tqdm

# üìÇ Lade 10 Mio Nullstellen aus Google Drive
from google.colab import drive
drive.mount('/content/drive')

# üìå Definiere den Dateipfad zu den Nullstellen
dateipfad = "/content/drive/MyDrive/zeros6.txt"

# üìå Lade Nullstellen aus Datei
def lade_nullstellen(dateipfad, max_anzahl=10_000_000):
    print(f"üìÇ Lade Nullstellen aus {dateipfad} ...")
    with open(dateipfad, 'r') as f:
        nullstellen = np.array([float(line.strip()) for line in f.readlines()[:max_anzahl]])
    print(f"‚úÖ {len(nullstellen)} Nullstellen erfolgreich geladen!")
    return nullstellen

# üéØ Berechne Abst√§nde der Nullstellen
def berechne_abstaende(nullstellen):
    return np.diff(nullstellen)

# üìå Freese-Formel Standard (FFS)
def freese_standard(n, A, B, C):
    return A * (np.sqrt(n)) + B * np.log(n) + C

# üìå Freese-Formel Oszillierend (FFO)
def freese_oszillierend(n, A, B, C, w, phi):
    return A * (np.sqrt(n)) + B * np.log(n) + C + np.sin(w * n + phi)

# üìå Fit-Funktion zur Optimierung
def optimiere_freese(nullstellen_abstaende, oszillierend=False):
    n_values = np.arange(1, len(nullstellen_abstaende) + 1)

    if oszillierend:
        params, _ = opt.curve_fit(freese_oszillierend, n_values, nullstellen_abstaende, p0=[1, 1, 1, 0.1, 0])
    else:
        params, _ = opt.curve_fit(freese_standard, n_values, nullstellen_abstaende, p0=[1, 1, 1])

    return params

# üìå Operator-Tests & Ableitungen mit SymPy
def operator_tests():
    n = sp.Symbol('n')
    A, B, C, w, phi = sp.symbols('A B C w phi')

    # Standardformel
    ffs = A * sp.sqrt(n) + B * sp.log(n) + C
    ffo = A * sp.sqrt(n) + B * sp.log(n) + C + sp.sin(w * n + phi)

    # Erste und zweite Ableitung
    ffs_diff1 = sp.diff(ffs, n)
    ffs_diff2 = sp.diff(ffs_diff1, n)
    ffo_diff1 = sp.diff(ffo, n)
    ffo_diff2 = sp.diff(ffo_diff1, n)

    print("\nüìä **Ableitungen der FFS & FFO:**")
    print(f"1. Ableitung (FFS): {ffs_diff1}")
    print(f"2. Ableitung (FFS): {ffs_diff2}")
    print(f"1. Ableitung (FFO): {ffo_diff1}")
    print(f"2. Ableitung (FFO): {ffo_diff2}")

# üìä Grafische Darstellung
def plot_freese(nullstellen_abstaende, params_ffs, params_ffo):
    n_values = np.arange(1, len(nullstellen_abstaende) + 1)

    plt.figure(figsize=(12, 6))

    # üìå Originaldaten plotten
    plt.plot(n_values, nullstellen_abstaende, 'k.', alpha=0.5, label="Originaldaten (Nullstellen-Abst√§nde)")

    # üìå Fit mit Standard-FFF
    plt.plot(n_values, freese_standard(n_values, *params_ffs), 'r-', label="Freese-Formel Standard (FFS)")

    # üìå Fit mit Oszillierender FFF
    plt.plot(n_values, freese_oszillierend(n_values, *params_ffo), 'b-', label="Freese-Formel Oszillierend (FFO)")

    plt.xlabel("n (Index der Nullstelle)")
    plt.ylabel("Abstand Œîn")
    plt.legend()
    plt.title("üìä Vergleich der FFS & FFO mit Originaldaten")
    plt.show()

# üìå Hauptprogramm
if __name__ == "__main__":
    # üìÇ Lade Nullstellen
    nullstellen = lade_nullstellen(dateipfad, max_anzahl=10_000_000)

    # üîç Berechne Nullstellen-Abst√§nde
    nullstellen_abstaende = berechne_abstaende(nullstellen)

    # üéØ Optimiere Freese-Formel Standard (FFS)
    params_ffs = optimiere_freese(nullstellen_abstaende, oszillierend=False)
    print(f"‚úÖ Optimierte FFS-Parameter: A={params_ffs[0]:.6f}, B={params_ffs[1]:.6f}, C={params_ffs[2]:.6f}")

    # üéØ Optimiere Freese-Formel Oszillierend (FFO)
    params_ffo = optimiere_freese(nullstellen_abstaende, oszillierend=True)
    print(f"üìå Optimierte FFO-Parameter: A={params_ffo[0]:.6f}, B={params_ffo[1]:.6f}, C={params_ffo[2]:.6f}, w={params_ffo[3]:.6f}, œÜ={params_ffo[4]:.6f}")

    # üî¨ Operator-Tests (Ableitungen etc.)
    operator_tests()

    # üìä Visualisierung der Ergebnisse
    plot_freese(nullstellen_abstaende, params_ffs, params_ffo)
```
