```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial

#  1 Datei laden
dateipfad = "/content/drive/MyDrive/beta_skala_v4.csv"  # Letzte stabile Version
df = pd.read_csv(dateipfad, header=None)

#  2 Daten extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values

#  3 Polynom 5. Grades anpassen
x = np.linspace(0, 1, len(beta_skala))  # Normierung des Index
p = Polynomial.fit(x, beta_skala, 5)  # Fit mit Grad 5

#  4 Korrektur anwenden
beta_skala_korrigiert = beta_skala / p(x)

#  5 Fehlerberechnung
fehler = beta_skala_korrigiert - 1.0
mittelwert_fehler = np.mean(np.abs(fehler))

#  6 Neue Datei speichern
datei_neu = "/content/drive/MyDrive/beta_skala_v5.csv"
df_neu = pd.DataFrame([zeta_nullstellen, beta_skala_korrigiert])
df_neu.to_csv(datei_neu, index=False, header=False)

#  7 Plot der neuen Skala
plt.figure(figsize=(12, 4))
plt.plot(beta_skala_korrigiert, label="Beta-Skala (5. Grad)", color="blue")
plt.axhline(1, color="gray", linestyle="dashed", label="Zielwert 1.0")
plt.xlabel("Index")
plt.ylabel("Wert")
plt.legend()
plt.title("Beta-Skala nach 5. Grad-Korrektur")
plt.show()

#  8 Fehlerplot
plt.figure(figsize=(12, 4))
plt.plot(fehler, label="Fehler (Beta - 1.0)", color="red")
plt.axhline(0, color="gray", linestyle="dashed")
plt.xlabel("Index")
plt.ylabel("Fehler")
plt.legend()
plt.title("Fehlerverlauf nach 5. Grad-Korrektur")
plt.show()

print(f"\n Mittelwert des Fehlers: {mittelwert_fehler}")
print(f" Korrigierte Datei gespeichert: {datei_neu}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft, fftfreq

#  1 Datei laden (Basis v5!)
dateipfad = "/content/drive/MyDrive/beta_skala_v5.csv"  # Falls ntig, Pfad anpassen!
df = pd.read_csv(dateipfad, header=None)

#  2 Daten extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values
index = np.arange(len(beta_skala))

#  3 Polynom 6. Grades zur Korrektur definieren
def poly6(x, a, b, c, d, e, f, g):
    return a*x**6 + b*x**5 + c*x**4 + d*x**3 + e*x**2 + f*x + g

#  4 Curve Fitting (Fehlermodell)
popt, _ = curve_fit(poly6, index, beta_skala)

#  5 Neue Beta-Skala berechnen
beta_skala_korrigiert = beta_skala / poly6(index, *popt)

#  6 Fehlerreihe bestimmen
fehler = beta_skala_korrigiert - 1.0

#  7 Neue Datei speichern (v6)
df_neu = pd.DataFrame([zeta_nullstellen, beta_skala_korrigiert])
datei_neu = "/content/drive/MyDrive/beta_skala_v6.csv"
df_neu.to_csv(datei_neu, index=False, header=False)

#  8 Fehlerverlauf plotten
plt.figure(figsize=(12, 4))
plt.plot(index, fehler, color="red", label="Fehler (Beta - 1.0)")
plt.axhline(0, color="gray", linestyle="dashed")
plt.xlabel("Index")
plt.ylabel("Fehler")
plt.legend()
plt.title("Fehlerverlauf nach 6. Grad-Korrektur")
plt.show()

#  Mittelwert des Fehlers berechnen
mittelwert_fehler = np.mean(np.abs(fehler))
print(f" Mittelwert des Fehlers: {mittelwert_fehler}")

#  9 Fourier-Analyse der Fehlerreihe
N = len(fehler)
fehler_spektrum = np.abs(fft(fehler))[:N // 2]
frequenzen = fftfreq(N)[:N // 2]

#  Dominante Frequenzen finden
dom_freqs = frequenzen[np.argsort(-fehler_spektrum)[:5]]
dom_amplituden = np.sort(-fehler_spektrum)[:5] * -1

#    Spektrum plotten
plt.figure(figsize=(12, 4))
plt.plot(frequenzen, fehler_spektrum, color="brown", alpha=0.8, label="Fourier-Spektrum des Fehlers")
plt.scatter(dom_freqs, dom_amplituden, color="blue", label="Dominante Frequenzen")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.legend()
plt.title("Fourier-Spektrum der Fehlerreihe")
plt.show()

#  Ergebnis ausgeben
print(f" Dominante Frequenzen: {dom_freqs}")
print(f" Korrespondierende Amplituden: {dom_amplituden}")
print(f"\n Korrigierte Datei gespeichert: {datei_neu}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

#  1 Pfad zur Beta-Skala-Datei
dateipfad = "/content/drive/MyDrive/beta_skala_v5.csv"

#  2 CSV-Datei laden
df = pd.read_csv(dateipfad, header=None)
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values

#  3 Fehlerfunktion berechnen (Beta-Skala - 1.0)
fehler = beta_skala - 1.0
index = np.arange(len(fehler))

#  4 Mittelwert anzeigen
mittelwert = np.mean(np.abs(fehler))
print(f" Mittelwert des Fehlers: {mittelwert}")

#  5 Fourier-Analyse
N = len(fehler)
fft_spektrum = fft(fehler)
frequenzen = fftfreq(N, d=1)

# Nur positive Frequenzen & Amplituden extrahieren
frequenzen_pos = frequenzen[:N // 2]
amplituden_pos = 2.0 / N * np.abs(fft_spektrum[:N // 2])

#  6 Dominante Frequenzen extrahieren
anzahl_dominant = 5
dominante_indices = np.argsort(amplituden_pos)[-anzahl_dominant:][::-1]
dominante_freqs = frequenzen_pos[dominante_indices]
dominante_ampls = amplituden_pos[dominante_indices]

print(" Dominante Frequenzen:", dominante_freqs)
print(" Korrespondierende Amplituden:", dominante_ampls)

#  7 Visualisierung des Spektrums
plt.figure(figsize=(12, 4))
plt.plot(frequenzen_pos, amplituden_pos, color='darkblue')
plt.scatter(dominante_freqs, dominante_ampls, color='red', label='Dominant')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Fehlerreihe")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  8 Fourier-Daten speichern (optional)
fourier_df = pd.DataFrame({
    "Frequenz": frequenzen_pos,
    "Amplitude": amplituden_pos
})
fourier_df.to_csv("/content/drive/MyDrive/beta_skala_v5_fourier.csv", index=False)
print(" Fourier-Daten gespeichert unter: beta_skala_v5_fourier.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
from scipy.fft import fft

#  1 Datei laden (z.B. korrigierte Skala v5)
dateipfad = "/content/drive/MyDrive/beta_skala_v5.csv"
df = pd.read_csv(dateipfad, header=None)

#  2 Daten extrahieren
zeta_nullstellen = df.iloc[0, :].values
beta_skala = df.iloc[1, :].values
index = np.arange(len(beta_skala))

#  3 Fehler berechnen zur Zielskala 1.0
fehler = beta_skala - 1.0
mittelwert_fehler = np.mean(np.abs(fehler))
print(f" Mittelwert des Fehlers: {mittelwert_fehler}")

#  4 Fourier-Analyse
N = len(fehler)
frequenzen = np.fft.fftfreq(N, d=1)
spektrum = fft(fehler)
amplituden = 2.0 / N * np.abs(spektrum)

#  5 Dominante Frequenzen finden
anzahl = 5
indices = np.argsort(amplituden)[-anzahl:][::-1]
dominante_freqs = frequenzen[indices]
dominante_amplituden = amplituden[indices]

print(f"\n Dominante Frequenzen: {dominante_freqs}")
print(f" Korrespondierende Amplituden: {dominante_amplituden}")

#  6 Symbolische Darstellung als Brche
print("\n Symbolische Darstellung der dominanten Frequenzen:")
for f in dominante_freqs:
    bruch = Fraction(f).limit_denominator(100000)
    print(f"Frequenz: {f:.10f}    {bruch.numerator}/{bruch.denominator}")

#  7 Fehlerverlauf plotten
plt.figure(figsize=(12, 4))
plt.plot(index, fehler, color="darkred", label="Fehler (Beta - 1.0)", alpha=0.7)
plt.axhline(0, color="gray", linestyle="dashed")
plt.title("Fehlerverlauf der Beta-Skala")
plt.xlabel("Index")
plt.ylabel("Fehler")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import cupy as cp
import matplotlib.pyplot as plt

# Beispiel-Daten: Fehler mit dominanter Frequenz
N = 2_000_000
x = cp.arange(N)
frequenz = 1 / 33300
signal = cp.sin(2 * cp.pi * frequenz * x) + 0.05 * cp.random.randn(N)

# Haar-Wavelet-Transformation (einfach, GPU-basiert)
def haar_wavelet_transform(data, level=1):
    result = data.copy()
    for _ in range(level):
        n = result.size // 2
        avg = (result[0::2] + result[1::2]) / cp.sqrt(2)
        diff = (result[0::2] - result[1::2]) / cp.sqrt(2)
        result = cp.concatenate((avg, diff))
    return result

# Transformation anwenden
wavelet_coeffs = haar_wavelet_transform(signal, level=1)

# Ergebnisse plotten (auf CPU kopieren)
plt.figure(figsize=(12, 4))
plt.plot(cp.asnumpy(wavelet_coeffs[:1000]), label="Haar-Wavelet-Koeffizienten (GPU)")
plt.title("GPU-basierte Haar-Wavelet-Analyse (1 Level)")
plt.xlabel("Index")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pywt

# === 1. Daten laden ===
pfad_v5 = "/content/drive/MyDrive/beta_skala_v5.csv"
df_v5 = pd.read_csv(pfad_v5, header=None)
beta_v5 = df_v5.iloc[1, :].values.astype(np.float64)
index = np.arange(len(beta_v5))
fehler = beta_v5 - 1.0

# === 2. Fehler reduzieren (Subset fr stabilen Fit) ===
x_fit = index[1:50000]
y_fit = fehler[1:50000]

# === 3. Modell definieren (Overflow-safe) ===
def model(x, A, w, phi, B, alpha):
    eps = 1e-3
    return A * np.sin(w * x + phi) + B / np.power(x + 1 + eps, alpha)

# === 4. Startparameter & Fit (mit hherer maxfev) ===
p0 = [1e-4, 2e-6, 0, 1e-4, 0.5]
bounds = ([0, 0, -np.pi, 0, 0], [1, 1e-2, np.pi, 1, 5])
popt, _ = curve_fit(model, x_fit, y_fit, p0=p0, bounds=bounds, maxfev=10000)

# === 5. Plot Originalfehler & Fit ===
plt.figure(figsize=(12, 4))
plt.plot(x_fit, y_fit, label="Fehler (Original)", color="red", alpha=0.6)
plt.plot(x_fit, model(x_fit, *popt), label="Modell-Fit", color="green")
plt.title("Symbolischer Fit fr Beta-Fehler v5")
plt.xlabel("Index")
plt.ylabel("Fehler")
plt.legend()
plt.grid(True)
plt.show()

# === 6. Fit-Parameter anzeigen ===
param_names = ["A (Amplitude)", " (Frequenz)", " (Phase)", "B (Skalenfaktor)", " (Exponent)"]
for name, val in zip(param_names, popt):
    print(f"{name}: {val:.10f}")

# === 7. Wavelet-Analyse auf Fehlerreihe ===
scales = np.arange(1, 256)
coeffs, _ = pywt.cwt(fehler[:2_000_000], scales, 'morl')

plt.figure(figsize=(12, 5))
plt.imshow(np.abs(coeffs), extent=[0, 2_000_000, 1, 256], cmap='jet', aspect='auto', vmax=0.001)
plt.colorbar(label="Amplitude")
plt.yscale('log')
plt.xlabel("Index")
plt.ylabel("Frequenz")
plt.title("Wavelet-Spektrum der Fehlerwerte (Beta-Skala v5)")
plt.show()
```

```python
#  GPU-optimierte Analyse der Beta-Fehlerreihe v5 mit Fit-Korrektur
import cupy as cp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#  1 Daten laden
pfad_v5 = "/content/drive/MyDrive/beta_skala_v5.csv"
df_v5 = pd.read_csv(pfad_v5, header=None, low_memory=False).astype(float)

#  2 Daten extrahieren
index = cp.arange(len(df_v5.columns))  # Index als GPU-Array
beta_v5 = cp.asarray(df_v5.iloc[1, :].values)  # Beta-Werte auf GPU
fehler = beta_v5 - 1.0  # Fehlerberechnung auf GPU

#  3 Fit-Modell definieren
def model_fehler(x, A, w, phi, B, alpha):
    return A * cp.sin(w * x + phi) + B / cp.power(x + 1, alpha)

#  4 Curve Fit fr gesamte Fehlerreihe
popt, _ = curve_fit(lambda x, A, w, phi, B, alpha: cp.asnumpy(model_fehler(cp.asarray(x), A, w, phi, B, alpha)),
                    cp.asnumpy(index), cp.asnumpy(fehler),
                    p0=[1e-4, 2e-6, 0, 1e-4, 0.5])

#  5 Fit-Parameter ausgeben
A, w, phi, B, alpha = popt
print(f"A (Amplitude): {A}")
print(f" (Frequenz): {w}")
print(f" (Phase): {phi}")
print(f"B (Skalenfaktor): {B}")
print(f" (Exponent): {alpha}")

#  6 Fehlerverlauf & Fit-Approximation plotten
plt.figure(figsize=(12, 4))
plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(fehler[:50000]), color="red", alpha=0.5, label="Fehler (Original)")
plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(model_fehler(index[:50000], *popt)), color="green", alpha=0.8, label="Fit-Modell")
plt.axhline(0, color="gray", linestyle="dashed")
plt.legend()
plt.title("Fehlerverlauf der Beta-Skala v5 mit Fit-Modell")
plt.show()

#  7 Korrektur der Beta-Skala mit Fit-Modell
beta_korrigiert = beta_v5 - model_fehler(index, *popt)
fehler_korrigiert = beta_korrigiert - 1.0

#  8 Fehlerverlauf nach Korrektur plotten
plt.figure(figsize=(12, 4))
plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(fehler_korrigiert[:50000]), color="blue", alpha=0.8, label="Fehler nach Korrektur")
plt.axhline(0, color="gray", linestyle="dashed")
plt.legend()
plt.title("Fehlerverlauf nach Fit-Korrektur")
plt.show()

#  9 Korrigierte Beta-Skala speichern
df_korrigiert = pd.DataFrame(cp.asnumpy(beta_korrigiert))
df_korrigiert.to_csv("/content/drive/MyDrive/beta_skala_v5_korrigiert.csv", index=False, header=False)
print(" Korrigierte Datei gespeichert: beta_skala_v5_korrigiert.csv")
```

```python
#  GPU-optimierte Analyse der Beta-Fehlerreihe v5 mit Fit-Korrektur
import cupy as cp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#  1 Daten laden
pfad_v5 = "/content/drive/MyDrive/beta_skala_v5.csv"
df_v5 = pd.read_csv(pfad_v5, header=None, low_memory=False).astype(float)

#  2 Daten extrahieren
index = cp.arange(len(df_v5.columns))  # Index als GPU-Array
beta_v5 = cp.asarray(df_v5.iloc[1, :].values)  # Beta-Werte auf GPU
fehler = beta_v5 - 1.0  # Fehlerberechnung auf GPU

#  3 Fit-Modell definieren (angepasst)
def model_fehler(x, A, w, phi, B, alpha):
    return A * cp.sin(w * x + phi) + B / (x + 1) ** alpha  # Robustere Potenzfunktion

#  4 Curve Fit mit besseren Startwerten & maxfev erhhen
try:
    popt, _ = curve_fit(lambda x, A, w, phi, B, alpha: cp.asnumpy(model_fehler(cp.asarray(x), A, w, phi, B, alpha)),
                        cp.asnumpy(index[:50000]), cp.asnumpy(fehler[:50000]),  #  Nur erste 50.000 Werte!
                        p0=[0.97, 1e-8, -1.57, 0.97, 0.0006],  #  Bessere Startwerte
                        maxfev=10000)  #  Hhere Anzahl erlaubter Funktionsaufrufe

    #  5 Fit-Parameter ausgeben
    A, w, phi, B, alpha = popt
    print(f"A (Amplitude): {A}")
    print(f" (Frequenz): {w}")
    print(f" (Phase): {phi}")
    print(f"B (Skalenfaktor): {B}")
    print(f" (Exponent): {alpha}")

    #  6 Fehlerverlauf & Fit-Approximation plotten
    plt.figure(figsize=(12, 4))
    plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(fehler[:50000]), color="red", alpha=0.5, label="Fehler (Original)")
    plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(model_fehler(index[:50000], *popt)), color="green", alpha=0.8, label="Fit-Modell")
    plt.axhline(0, color="gray", linestyle="dashed")
    plt.legend()
    plt.title("Fehlerverlauf der Beta-Skala v5 mit Fit-Modell")
    plt.show()

    #  7 Korrektur der Beta-Skala mit Fit-Modell
    beta_korrigiert = beta_v5 - model_fehler(index, *popt)
    fehler_korrigiert = beta_korrigiert - 1.0

    #  8 Fehlerverlauf nach Korrektur plotten
    plt.figure(figsize=(12, 4))
    plt.plot(cp.asnumpy(index[:50000]), cp.asnumpy(fehler_korrigiert[:50000]), color="blue", alpha=0.8, label="Fehler nach Korrektur")
    plt.axhline(0, color="gray", linestyle="dashed")
    plt.legend()
    plt.title("Fehlerverlauf nach Fit-Korrektur")
    plt.show()

    #  9 Korrigierte Beta-Skala speichern
    df_korrigiert = pd.DataFrame(cp.asnumpy(beta_korrigiert))
    df_korrigiert.to_csv("/content/drive/MyDrive/beta_skala_v5_korrigiert.csv", index=False, header=False)
    print(" Korrigierte Datei gespeichert: beta_skala_v5_korrigiert.csv")

except RuntimeError as e:
    print(" Curve Fit ist nicht konvergiert. Fehler:", e)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# === 1. Verschiedene Beta-Werte definieren ===
beta_werte = {
    "1": 1,
    "1/2": 1/2,
    "1/33": 1/33,
    "1/137": 1/137,
    "7/33300": 7/33300,
    "0.0001927701": 0.0001927701,
    "ln(3)": np.log(3)
}

# === 2. Berechnung der komplexen Punkte H = -1 / e^{i} ===
punkte = {}
for name, beta in beta_werte.items():
    h = -1 / np.exp(1j * np.pi * beta)
    punkte[name] = h

# === 3. Plot im Einheitskreis ===
fig, ax = plt.subplots(figsize=(6,6))
theta = np.linspace(0, 2*np.pi, 1000)
kreis = np.exp(1j * theta)
ax.plot(kreis.real, kreis.imag, 'gray', linestyle="--", label="Einheitskreis")

# === 4. Punkte plotten ===
for name, h in punkte.items():
    ax.plot(h.real, h.imag, 'o', label=f" = {name}")
    ax.plot([0, h.real], [0, h.imag], linestyle='-', alpha=0.6)

ax.set_aspect("equal")
ax.set_title("Euler-Freese-Komplexpunkte: $H = -1 / e^{i}$")
ax.set_xlabel("Re")
ax.set_ylabel("Im")
ax.grid(True)
ax.legend()
plt.show()

# === 5. Ausgabe der Koordinaten + Betrge und Phasen ===
print("Komplexe Darstellung der Euler-Freese-Identitt:")
for name, h in punkte.items():
    betrag = np.abs(h)
    phase = np.angle(h)
    print(f" = {name:<15}  H = {h.real:+.6f}{h.imag:+.6f}j | Betrag = {betrag:.6f} | Phase = {phase:.6f}")
```
