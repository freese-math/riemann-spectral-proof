```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
import time

# === 1. Wertebereich für n
n_values = np.arange(2, 1000)

# === 2. Zielfunktion (Optimierung des Grenzwerts von exp(iπβ(n)) + 1)
def objective(params):
    A, a, b, f, phi, c, mu = params
    log_term = np.log(n_values + mu)
    beta_n = A + (a / log_term) + b * np.cos(2 * np.pi * f * log_term + phi) + c * n_values
    result = np.exp(1j * np.pi * beta_n) + 1
    weighted_error = (np.abs(result)**2) * (n_values / n_values.max())
    return np.mean(weighted_error)

# === 3. Parametergrenzen
bounds = [
    (-10, 10),         # A
    (-10, 10),         # a
    (-10, 10),         # b
    (0, 1.0),          # f
    (0, 2*np.pi),      # phi
    (-1e-3, 1e-3),     # c
    (0.1, 10.0)        # mu
]

# === 4. Fortschritt-Callback
def callback_fn(xk, convergence):
    print(f"Aktuelles Ziel: {objective(xk):.6e}, Konvergenz: {convergence:.6f}")

# === 5. Zeitmessung
start_time = time.time()

# === 6. Optimierung
result = differential_evolution(
    objective,
    bounds,
    callback=callback_fn,
    strategy='best1bin',
    maxiter=300,
    popsize=15,
    tol=1e-8,
    disp=True
)

# === 7. Ergebnisse
end_time = time.time()
optimal_params = result.x
min_error = result.fun

# === 8. Rekonstruktion
A, a, b, f, phi, c, mu = optimal_params
log_term = np.log(n_values + mu)
beta_n = A + (a / log_term) + b * np.cos(2 * np.pi * f * log_term + phi) + c * n_values
epsilon = np.exp(1j * np.pi * beta_n) + 1

# === 9. Ausgabe
print("\nOptimale Parameter:")
param_names = ['A', 'a', 'b', 'f', 'phi', 'c', 'mu']
for name, val in zip(param_names, optimal_params):
    print(f"{name:>4} = {val:.6g}")
print(f"\nMinimierter Fehler (RMSE): {min_error:.2e}")
print(f"Benötigte Zeit: {end_time - start_time:.2f} Sekunden")

# === 10. Plot
plt.figure(figsize=(8, 4))
plt.plot(n_values, np.abs(epsilon), label='|ε(n)|')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Residual |ε(n)| = |exp(iπβ(n)) + 1|")
plt.xlabel("n")
plt.ylabel("|ε(n)|")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Werte für Beta
beta_values = {
    "β = 1": 1,
    "β = 1/2": 1/2,
    "β = 1/33": 1/33,
    "β = 1/137": 1/137,
    "β = 7/33300": 7/33300,
    "β = 0.0001927701": 0.0001927701,
    "β = ln(3)": np.log(3)
}

# Funktion: H = -1 / exp(i π β)
def compute_H(beta):
    return -1 / np.exp(1j * np.pi * beta)

# Einheitskreis vorbereiten
circle = np.exp(1j * np.linspace(0, 2 * np.pi, 500))

# Plot
plt.figure(figsize=(6, 6))
plt.plot(circle.real, circle.imag, 'k--', alpha=0.3, label='Einheitskreis')

# Punkte und Linien zeichnen
for label, beta in beta_values.items():
    H = compute_H(beta)
    plt.plot(H.real, H.imag, 'o', label=label)
    plt.plot([0, H.real], [0, H.imag], alpha=0.5)

# Plot-Einstellungen
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.gca().set_aspect('equal')
plt.xlabel("Re")
plt.ylabel("Im")
plt.title("Euler-Freese-Komplexpunkte: H = -1 / exp(iπβ)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle=':')
plt.tight_layout()
plt.show()
```
