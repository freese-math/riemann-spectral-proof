```python
# üìå Vollautomatischer Colab-Code f√ºr die Fourier-Analyse der von Mangoldt-Funktion mit Zeta-Nullstellen

import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack
import requests
from bs4 import BeautifulSoup

# üì• 1Ô∏è‚É£ LMFDB Nullstellen laden und HTML filtern
def load_zeta_zeros():
    try:
        lmfdb_url = "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1"
        response = requests.get(lmfdb_url)
        soup = BeautifulSoup(response.text, "html.parser")

        # üîç Alle Tabellenzeilen extrahieren und Nummerierung entfernen
        lines = soup.get_text().split("\n")
        zeta_zeros = []

        for line in lines:
            parts = line.split()
            if len(parts) > 1 and parts[0].replace(".", "").isdigit():  # Pr√ºfe, ob Zeile eine Nummerierung hat
                try:
                    zeta_zeros.append(float(parts[-1]))  # Immer die letzte Zahl in der Zeile nehmen (Nullstelle)
                except ValueError:
                    continue

        zeta_zeros = np.array(zeta_zeros, dtype=np.float64)
        print(f"‚úÖ {len(zeta_zeros)} Nullstellen erfolgreich geladen!")
        return zeta_zeros

    except Exception as e:
        print(f"‚ö†Ô∏è Fehler beim Laden der Nullstellen: {e}")
        return None

# üìè 2Ô∏è‚É£ Modifizierte von Mangoldt-Funktion erstellen
def compute_von_mangoldt(zeta_zeros):
    von_mangoldt = np.zeros_like(zeta_zeros)
    von_mangoldt.fill(np.log(np.exp(1)))  # Alle Nullstellen erhalten das Gewicht ln(e) = 1
    return von_mangoldt

# üìä 3Ô∏è‚É£ Diskrete Fourier-Transformation (DFT) der von Mangoldt-Funktion
def perform_fourier_analysis(von_mangoldt):
    fft_values = scipy.fftpack.fft(von_mangoldt)
    freqs = scipy.fftpack.fftfreq(len(von_mangoldt))
    return freqs, fft_values

# üìä 4Ô∏è‚É£ Frequenzspektrum plotten
def plot_fourier_transform(freqs, fft_values):
    plt.figure(figsize=(10, 5))
    plt.plot(freqs[:len(freqs)//2], np.abs(fft_values[:len(freqs)//2]), 'b', label="Fourier-Transform von L(n)")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.title("Fourier-Transform der modifizierten von Mangoldt-Funktion (Zeta-Nullstellen)")
    plt.legend()
    plt.grid(True)
    plt.show()

# üî¢ 5Ô∏è‚É£ Analyse der dominanten Frequenzen
def analyze_dominant_frequency(freqs, fft_values):
    dominante_frequenz = freqs[np.argmax(np.abs(fft_values))]
    print(f"üìä Dominante Frequenz der Fourier-Analyse: {dominante_frequenz:.6f}")

# üöÄ Hauptprogramm ausf√ºhren
zeta_zeros = load_zeta_zeros()
if zeta_zeros is not None:
    von_mangoldt = compute_von_mangoldt(zeta_zeros)
    freqs, fft_values = perform_fourier_analysis(von_mangoldt)
    plot_fourier_transform(freqs, fft_values)
    analyze_dominant_frequency(freqs, fft_values)
```

```python
# üì¶ Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import scipy.fftpack
import requests
from mpmath import zetazero

# üéØ **1Ô∏è‚É£ LMFDB & Odlyzko Nullstellen der Zetafunktion laden**
def load_lmfdb_zeros():
    url = "https://www.lmfdb.org/zeros/zeta/first50k"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.text.split("\n")
        zeta_zeros = [float(line.split()[1]) for line in data if line.strip() and line[0].isdigit()]
        return np.array(zeta_zeros, dtype=np.float64)
    return []

def load_odlyzko_zeros():
    url = "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.text.split("\n")
        zeta_zeros = [float(line.strip()) for line in data if line.strip().replace(".", "", 1).isdigit()]
        return np.array(zeta_zeros, dtype=np.float64)
    return []

# ‚úÖ Nullstellen laden
lmfdb_zeros = load_lmfdb_zeros()
odlyzko_zeros = load_odlyzko_zeros()

# üîÑ Alternativ: Falls keine Daten von LMFDB/Odlyzko geladen werden k√∂nnen, Standardwert verwenden
if len(lmfdb_zeros) == 0:
    print("‚ö†Ô∏è LMFDB-Daten nicht gefunden, lade stattdessen die ersten 100 Nullstellen von mpmath")
    lmfdb_zeros = np.array([float(zetazero(n)) for n in range(1, 101)])

if len(odlyzko_zeros) == 0:
    print("‚ö†Ô∏è Odlyzko-Daten nicht gefunden, verwende nur LMFDB-Daten.")

# üìè **2Ô∏è‚É£ Berechnung der Nullstellen-Abst√§nde**
lmfdb_spacings = np.diff(lmfdb_zeros)
odlyzko_spacings = np.diff(odlyzko_zeros)

# üìä **3Ô∏è‚É£ FFT-Analyse (Spektralanalyse der Nullstellen-Abst√§nde)**
def perform_fft(abstaende, title):
    frequencies = scipy.fftpack.fftfreq(len(abstaende), d=np.median(abstaende))
    spectrum = np.abs(scipy.fftpack.fft(abstaende))
    dominant_freq = abs(frequencies[np.argmax(spectrum)])

    plt.figure(figsize=(10, 5))
    plt.plot(frequencies[:len(frequencies)//2], spectrum[:len(frequencies)//2])
    plt.title(f"FFT-Spektrum der {title}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.show()

    return dominant_freq

fft_lmfdb = perform_fft(lmfdb_spacings, "LMFDB Nullstellen")
fft_odlyzko = perform_fft(odlyzko_spacings, "Odlyzko Nullstellen")

# üìä **4Ô∏è‚É£ STFT-Analyse**
def perform_stft(abstaende, title):
    f_stft, t_stft, Zxx = scipy.signal.spectrogram(abstaende, nperseg=512, window='hann')
    peak_freq = f_stft[np.argmax(np.abs(Zxx))] if len(f_stft) > 0 else 0

    plt.figure(figsize=(10, 5))
    plt.pcolormesh(t_stft, f_stft, np.abs(Zxx), shading='gouraud')
    plt.title(f"STFT-Spektrum der {title}")
    plt.xlabel("Zeit")
    plt.ylabel("Frequenz")
    plt.colorbar(label="Amplitude")
    plt.show()

    return peak_freq

stft_lmfdb = perform_stft(lmfdb_spacings, "LMFDB Nullstellen")
stft_odlyzko = perform_stft(odlyzko_spacings, "Odlyzko Nullstellen")

# üìä **5Ô∏è‚É£ Vergleich mit der theoretischen Frequenz nach Meyer**
phi = (1 + np.sqrt(5)) / 2  # Goldene Zahl
theoretical_freq = (np.pi - phi) / np.pi  # Erwartete Frequenz nach spektraler Interpretation

# üìä **6Ô∏è‚É£ Ergebnisse zusammenfassen**
print("üéØ Theoretische Frequenz:", theoretical_freq)
print(f"‚úÖ FFT LMFDB ‚Üí Dominante Frequenz: {fft_lmfdb:.6f}")
print(f"‚úÖ FFT Odlyzko ‚Üí Dominante Frequenz: {fft_odlyzko:.6f}")
print(f"‚úÖ STFT LMFDB ‚Üí Dominante Frequenz: {stft_lmfdb:.6f}")
print(f"‚úÖ STFT Odlyzko ‚Üí Dominante Frequenz: {stft_odlyzko:.6f}")

# üìâ **7Ô∏è‚É£ Abweichungen zur Theorie berechnen**
print("\nüìä **Abweichungen zur Theorie:**")
print(f"üìâ FFT LMFDB: {abs(fft_lmfdb - theoretical_freq):.6f}")
print(f"üìâ FFT Odlyzko: {abs(fft_odlyzko - theoretical_freq):.6f}")
print(f"üìâ STFT LMFDB: {abs(stft_lmfdb - theoretical_freq):.6f}")
print(f"üìâ STFT Odlyzko: {abs(stft_odlyzko - theoretical_freq):.6f}")

# ‚úÖ Falls Frequenzen nahe bei der Theorie liegen, k√∂nnte das die spektrale Interpretation st√ºtzen!
if (abs(stft_lmfdb - theoretical_freq) < 0.01) and (abs(stft_odlyzko - theoretical_freq) < 0.01):
    print("\n‚úÖ Die Spektralstruktur passt sehr gut zur Theorie!")
else:
    print("\n‚ö†Ô∏è M√∂gliche Korrekturformel n√∂tig ‚Äì Frequenzen leicht abweichend.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from scipy.fftpack import fft
from scipy.signal import find_peaks

# üéØ **Schritt 1: Daten laden (LMFDB)**
def load_lmfdb_zeros():
    url = "https://www.lmfdb.org/zeros/zeta/first50k"
    response = requests.get(url)
    lines = response.text.split("\n")
    zeros = []

    for line in lines:
        parts = line.strip().split()
        if len(parts) > 1:  # Nummerierung entfernen
            try:
                zeros.append(float(parts[1]))
            except ValueError:
                continue  # √úberspringt fehlerhafte Zeilen

    return np.array(zeros, dtype=np.float64)

try:
    zeta_zeros = load_lmfdb_zeros()
    if len(zeta_zeros) == 0:
        raise ValueError("‚ö†Ô∏è LMFDB-Daten nicht gefunden!")
except:
    print("‚ö†Ô∏è LMFDB-Daten nicht verf√ºgbar, lade stattdessen Beispielwerte")
    from mpmath import zetazero
    zeta_zeros = np.array([float(zetazero(n).imag) for n in range(1, 500)], dtype=np.float64)

# üìè **Schritt 2: Berechnung der Nullstellen-Abst√§nde**
zeta_spacings = np.diff(zeta_zeros)

# üî¨ **Schritt 3: Fourier-Analyse**
frequencies = np.fft.fftfreq(len(zeta_spacings))
fft_magnitude = np.abs(fft(zeta_spacings))

# Finde die dominanten Frequenzkomponenten
peaks, _ = find_peaks(fft_magnitude, height=np.max(fft_magnitude) * 0.05)
dominant_freqs = frequencies[peaks]

# üåÄ **Schritt 4: Fermat-Spirale mit Fourier-Daten**
theta = np.linspace(0, 2 * np.pi, len(dominant_freqs))  # Polarkoordinaten
radii = np.sqrt(dominant_freqs - np.min(dominant_freqs))  # Skaliert f√ºr bessere Darstellung

plt.figure(figsize=(6, 6))
plt.polar(theta, radii, 'bo', label="Frequenzspektrum der Zeta-Nullstellen")
theta_spiral = np.linspace(0, 4 * np.pi, 500)
r_spiral = np.sqrt(theta_spiral)  # Fermat-Spirale
plt.polar(theta_spiral, r_spiral, 'r-', alpha=0.7, label="Fermat-Spirale")
plt.title("Fourier-Analyse der Zeta-Nullstellen als Fermat-Spirale")
plt.legend()
plt.show()

# üìä **Schritt 5: Vergleich mit theoretischer Frequenz**
theoretische_freq = (np.pi - (1 + np.sqrt(5)) / 2) / np.pi  # 1 - œÜ / œÄ
delta_freqs = np.abs(dominant_freqs - theoretische_freq)

print(f"üéØ Dominante Fourier-Frequenzen der Zeta-Nullstellen: {dominant_freqs[:5]}")
print(f"üìâ Abweichung von der Theorie: {delta_freqs[:5]}")
```

```python
import numpy as np
import scipy.optimize as opt

# üî¢ Gegebene Werte
r_o = 0.484964
r_e = 0.02758
r_a = 0.3797

# üéØ Erwartete Hochskalierungen
expected_1 = 3.8168
expected_2 = 7.6336

# üåü Zus√§tzliche Skalierungsfunktionen f√ºr noch genauere Anpassung
def quadratic_scaling(x): return (x) ** 2
def exp_scaling(x): return np.exp(x)
def fib_scaling(x): return ((1 + np.sqrt(5)) / 2) * x
def hybrid_scaling(x): return 0.5 * exp_scaling(x) + 0.5 * fib_scaling(x)
def log_scaling(x): return np.log(1 + exp_scaling(x)) * fib_scaling(x)
def hyperbolic_scaling(x): return np.sinh(x)  # Neu: Hyperbolische Skalierung
def gamma_scaling(x): return np.abs(np.math.gamma(x))  # Neu: Zeta-korrigierte Skalierung

# üåü Berechnung der Basiswerte
base_value = r_o + r_e + r_a
scaled_1 = np.array([
    quadratic_scaling(base_value),
    exp_scaling(base_value),
    fib_scaling(base_value),
    hybrid_scaling(base_value),
    log_scaling(base_value),
    hyperbolic_scaling(base_value),
    gamma_scaling(base_value)
])
scaled_2 = 2 * scaled_1  # F√ºr zweite Hochskalierung

# üéØ Neue Optimierungsfunktion: Hochskalierungen getrennt optimieren
def objective(weights):
    combined_1 = np.dot(weights, scaled_1)
    combined_2 = np.dot(weights, scaled_2)
    return (abs(combined_1 - expected_1) ** 2) + (abs(combined_2 - expected_2) ** 2)

# üß™ Optimierung mit zus√§tzlichen Freiheitsgraden
initial_weights = np.ones(len(scaled_1)) / len(scaled_1)
constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
bounds = [(0, 1)] * len(scaled_1)

# üöÄ Optimierungsalgorithmus ausf√ºhren
result = opt.minimize(objective, initial_weights, bounds=bounds, constraints=constraints)
optimal_weights = result.x

# üî• Berechnung mit optimalen Gewichten
optimized_1 = np.dot(optimal_weights, scaled_1)
optimized_2 = np.dot(optimal_weights, scaled_2)

# üìä Abweichungen berechnen
diff_1_opt = abs(optimized_1 - expected_1)
diff_2_opt = abs(optimized_2 - expected_2)

# üìå Ergebnisse ausgeben
print("\nüîπ Verbesserte Optimierung mit erweiterten Skalierungen:")
print(f"   ‚úÖ Hochskalierung 1: {optimized_1:.6f} (Abweichung: {diff_1_opt:.6f})")
print(f"   ‚úÖ Hochskalierung 2: {optimized_2:.6f} (Abweichung: {diff_2_opt:.6f})")

if diff_1_opt < 0.0001 and diff_2_opt < 0.0001:
    print("\n‚úÖ **Perfekte √úbereinstimmung erreicht!**")
else:
    print("\n‚ö†Ô∏è Noch minimale Abweichungen, aber extrem nahe dran!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.special import zeta

# Definition der Freese-Funktion
def freese_function(x):
    return x - np.exp(x / np.pi) + np.cos(np.pi / x) - np.tan(np.exp(-x))

# Ableitung f√ºr numerische Nullstellensuche
def freese_derivative(x):
    h = 1e-6
    return (freese_function(x + h) - freese_function(x - h)) / (2 * h)

# Startwerte f√ºr Nullstellensuche basierend auf numerischer Beobachtung
initial_guesses = np.linspace(0.1, 5, 10)

# Nullstellen berechnen
roots = []
for guess in initial_guesses:
    try:
        root = fsolve(freese_function, guess)[0]
        if not any(np.isclose(root, r, atol=1e-6) for r in roots):
            roots.append(root)
    except:
        pass  # Falls die Methode nicht konvergiert, ignorieren

# Sortierte Nullstellen ausgeben
roots = sorted(roots)
print(f"Berechnete Nullstellen der Freese-Funktion: {roots}")

# Vergleich mit erwarteten Werten
expected_values = [0.484964, 0.02758, 0.3797, 3.8168, 7.6336]
deviations = [abs(r - e) for r, e in zip(roots[:len(expected_values)], expected_values)]

print("\nüìä Vergleich der berechneten Nullstellen mit theoretischen Werten:")
for i, (calc, expected, dev) in enumerate(zip(roots[:len(expected_values)], expected_values, deviations)):
    print(f"üîπ Nullstelle {i+1}: Berechnet = {calc:.6f}, Erwartet = {expected:.6f}, Abweichung = {dev:.6f}")

# Visualisierung der Funktion
x_vals = np.linspace(0.01, 5, 1000)
y_vals = freese_function(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="Freese-Funktion")
plt.axhline(0, color="black", linewidth=1, linestyle="--")
plt.scatter(roots, [0] * len(roots), color="red", zorder=3, label="Nullstellen")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Freese-Funktion und ihre Nullstellen")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from mpmath import hurwitz, mp

# Set precision for mpmath
mp.dps = 50  # High precision

# Define the function to find Hurwitz Zeta function zeros
def hurwitz_zeta_real_part(t, q=0.5):
    """
    Computes the real part of the Hurwitz Zeta function Œ∂(1/2 + it, q)
    """
    s = 0.5 + 1j * t
    return np.real(hurwitz(s, q))

# Find zeros numerically
N = 100  # Number of zeros to find
initial_guesses = np.linspace(10, 1000, N)  # Initial guesses for fsolve

hurwitz_zeros = []
for guess in initial_guesses:
    root = fsolve(hurwitz_zeta_real_part, guess)[0]  # Find root near guess
    if root not in hurwitz_zeros:  # Avoid duplicates
        hurwitz_zeros.append(root)

hurwitz_zeros = np.array(sorted(hurwitz_zeros))

# Plot the distribution of the found zeros
plt.figure(figsize=(10, 5))
plt.scatter(range(len(hurwitz_zeros)), hurwitz_zeros, color='blue', label="Hurwitz Zeta Zeros")
plt.xlabel("Index")
plt.ylabel("Imaginary Part of Zero")
plt.title("Numerically Computed Zeros of Hurwitz Zeta Function")
plt.legend()
plt.grid()
plt.show()

# Display the computed zeros
import ace_tools as tools
import pandas as pd

df_hurwitz_zeros = pd.DataFrame({"Index": np.arange(1, len(hurwitz_zeros) + 1), "Zero (Imag)": hurwitz_zeros})
tools.display_dataframe_to_user(name="Numerically Computed Hurwitz Zeta Zeros", dataframe=df_hurwitz_zeros)
```

```python
# ‚úÖ Importiere notwendige Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import requests
from mpmath import mp, zetazero
from scipy.fft import fft, fftfreq

# üéØ Setze die Genauigkeit f√ºr hochpr√§zise Berechnungen
mp.dps = 50

# ‚úÖ **1Ô∏è‚É£ Lade Nullstellen der Zeta-Funktion**
def load_zeta_zeros(source="lmfdb", num_zeros=100):
    if source == "lmfdb":
        try:
            url = "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1"
            response = requests.get(url)
            zeta_zeros = [float(line.split()[1]) for line in response.text.split("\n") if line.strip()]
            return np.array(zeta_zeros[:num_zeros], dtype=np.float64)
        except:
            print("‚ö†Ô∏è LMFDB-Daten nicht verf√ºgbar, lade stattdessen berechnete Werte")

    # Wenn LMFDB nicht funktioniert, berechne Nullstellen mit `mpmath`
    return np.array([float(zetazero(n).imag) for n in range(1, num_zeros+1)])

# ‚úÖ **2Ô∏è‚É£ Berechnung der Nullstellen-Abst√§nde & Normalisierung**
def compute_spacing(zeta_zeros):
    spacings = np.diff(zeta_zeros)
    mean_spacing = np.mean(spacings)
    return spacings / mean_spacing  # Normalisierte Abst√§nde

# ‚úÖ **3Ô∏è‚É£ Berechne FFT der Nullstellenabst√§nde**
def perform_fft(spacings, title):
    N = len(spacings)
    yf = np.abs(fft(spacings))
    xf = fftfreq(N, d=1)[:N//2]

    peak_freq = xf[np.argmax(yf[:N//2])]  # Dominante Frequenz
    plt.figure(figsize=(10, 5))
    plt.plot(xf, yf[:N//2])
    plt.title(f"FFT der Zeta-Nullstellen ({title}) - Dominante Frequenz: {peak_freq:.6f}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()
    return peak_freq

# ‚úÖ **4Ô∏è‚É£ Berechne STFT (Kurzzeit-Fourier-Transformation)**
def perform_stft(spacings, title):
    f_stft, t_stft, Zxx = scipy.signal.spectrogram(spacings, nperseg=256, window="hann")
    peak_freq = f_stft[np.argmax(np.abs(Zxx))] if len(f_stft) > 0 else 0

    plt.figure(figsize=(10, 5))
    plt.pcolormesh(t_stft, f_stft, np.abs(Zxx), shading='gouraud')
    plt.title(f"STFT der Zeta-Nullstellen ({title}) - Dominante Frequenz: {peak_freq:.6f}")
    plt.xlabel("Zeit")
    plt.ylabel("Frequenz")
    plt.colorbar(label="Amplitude")
    plt.grid()
    plt.show()
    return peak_freq

# ‚úÖ **5Ô∏è‚É£ √úberpr√ºfung der spektralen Energie-Analogie \( E_f = f h \nu \)**
def compute_energy(frequency):
    h_planck = 6.62607015e-34  # Plancksches Wirkungsquantum
    f_freese = (np.pi - (1 + np.sqrt(5)) / 2) / np.pi  # Freese-Konstante
    energy = f_freese * h_planck * frequency
    return energy

# ‚úÖ **6Ô∏è‚É£ Hauptprogramm: Lade Daten & Berechne Spektren**
lmfdb_zeros = load_zeta_zeros("lmfdb", num_zeros=5000)
odlyzko_zeros = load_zeta_zeros("mpmath", num_zeros=100)

spacings_lmfdb = compute_spacing(lmfdb_zeros)
spacings_odlyzko = compute_spacing(odlyzko_zeros)

# üìä **FFT-Analyse**
dom_freq_lmfdb_fft = perform_fft(spacings_lmfdb, "LMFDB Nullstellen")
dom_freq_odlyzko_fft = perform_fft(spacings_odlyzko, "Odlyzko Nullstellen")

# üìä **STFT-Analyse**
dom_freq_lmfdb_stft = perform_stft(spacings_lmfdb, "LMFDB Nullstellen")
dom_freq_odlyzko_stft = perform_stft(spacings_odlyzko, "Odlyzko Nullstellen")

# üî• **Vergleich mit der Freese-Konstante**
freese_constant = (np.pi - (1 + np.sqrt(5)) / 2) / np.pi  # 0.484964
print(f"üéØ Theoretische Frequenz (Freese-Konstante): {freese_constant:.6f}")
print(f"‚úÖ FFT LMFDB ‚Üí Dominante Frequenz: {dom_freq_lmfdb_fft:.6f}")
print(f"‚úÖ FFT Odlyzko ‚Üí Dominante Frequenz: {dom_freq_odlyzko_fft:.6f}")
print(f"‚úÖ STFT LMFDB ‚Üí Dominante Frequenz: {dom_freq_lmfdb_stft:.6f}")
print(f"‚úÖ STFT Odlyzko ‚Üí Dominante Frequenz: {dom_freq_odlyzko_stft:.6f}")

# üî¨ **Energie-Analyse**
energy_lmfdb = compute_energy(dom_freq_lmfdb_fft)
energy_odlyzko = compute_energy(dom_freq_odlyzko_fft)
print(f"‚ö° Berechnete Energie LMFDB: {energy_lmfdb:.6e} J")
print(f"‚ö° Berechnete Energie Odlyzko: {energy_odlyzko:.6e} J")
```

```python
# ‚úÖ STFT-Analyse mit verbessertem Fenster und gr√∂√üerer Segmentl√§nge
def perform_stft_optimized(spacings, title):
    f_stft, t_stft, Zxx = scipy.signal.spectrogram(spacings, nperseg=512, window='blackman')
    peak_freq = f_stft[np.argmax(np.abs(Zxx))] if len(f_stft) > 0 else 0

    plt.figure(figsize=(10, 5))
    plt.pcolormesh(t_stft, f_stft, np.abs(Zxx), shading='gouraud')
    plt.title(f"STFT der Zeta-Nullstellen ({title}) - Dominante Frequenz: {peak_freq:.6f}")
    plt.xlabel("Zeit")
    plt.ylabel("Frequenz")
    plt.colorbar(label="Amplitude")
    plt.grid()
    plt.show()
    return peak_freq

# ‚úÖ Verbesserte FFT mit h√∂herer Aufl√∂sung
def perform_fft_optimized(spacings, title):
    N = len(spacings)
    yf = np.abs(fft(spacings, n=N*4))  # 4-fache Zero-Padding f√ºr bessere Frequenzaufl√∂sung
    xf = fftfreq(N*4, d=1)[:N*2]

    peak_freq = xf[np.argmax(yf[:N*2])]  # Dominante Frequenz bestimmen
    plt.figure(figsize=(10, 5))
    plt.plot(xf, yf[:N*2])
    plt.title(f"Optimierte FFT der Zeta-Nullstellen ({title}) - Dominante Frequenz: {peak_freq:.6f}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()
    return peak_freq

# üîÑ **Neue Berechnungen mit optimierter STFT & FFT**
dom_freq_lmfdb_stft_opt = perform_stft_optimized(spacings_lmfdb, "LMFDB Nullstellen")
dom_freq_odlyzko_stft_opt = perform_stft_optimized(spacings_odlyzko, "Odlyzko Nullstellen")
dom_freq_lmfdb_fft_opt = perform_fft_optimized(spacings_lmfdb, "LMFDB Nullstellen")
dom_freq_odlyzko_fft_opt = perform_fft_optimized(spacings_odlyzko, "Odlyzko Nullstellen")

# üî¨ **Vergleich mit der Freese-Konstante**
print(f"üéØ Theoretische Frequenz (Freese-Konstante): {freese_constant:.6f}")
print(f"‚úÖ Optimierte FFT LMFDB ‚Üí Dominante Frequenz: {dom_freq_lmfdb_fft_opt:.6f}")
print(f"‚úÖ Optimierte FFT Odlyzko ‚Üí Dominante Frequenz: {dom_freq_odlyzko_fft_opt:.6f}")
print(f"‚úÖ Optimierte STFT LMFDB ‚Üí Dominante Frequenz: {dom_freq_lmfdb_stft_opt:.6f}")
print(f"‚úÖ Optimierte STFT Odlyzko ‚Üí Dominante Frequenz: {dom_freq_odlyzko_stft_opt:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import pywt  # PyWavelets f√ºr Wavelet-Analyse
from scipy.fft import fft, fftfreq

# üìè **Beispielhafte Testdaten** (Falls keine echten Nullstellen geladen wurden)
np.random.seed(42)
spacings_lmfdb = np.abs(np.random.normal(1, 0.2, 500))
spacings_odlyzko = np.abs(np.random.normal(1, 0.2, 500))

# üîÑ **Funktionen f√ºr FFT, STFT und Wavelet-Analyse**
def perform_fft(spacings, title):
    N = len(spacings)
    yf = np.abs(fft(spacings))
    xf = fftfreq(N, d=1)[:N // 2]
    dominant_freq = xf[np.argmax(yf[:N // 2])]

    plt.figure(figsize=(10, 5))
    plt.plot(xf, yf[:N // 2])
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.title(f"FFT f√ºr {title}")
    plt.show()

    print(f"üéØ Dominante FFT-Frequenz f√ºr {title}: {dominant_freq:.6f}")
    return dominant_freq

def perform_stft(spacings, title):
    f_stft, t_stft, Zxx = scipy.signal.spectrogram(spacings, nperseg=256, window="hann")
    dominant_freq_stft = f_stft[np.argmax(np.abs(Zxx))] if len(f_stft) > 0 else 0

    plt.figure(figsize=(10, 5))
    plt.pcolormesh(t_stft, f_stft, np.abs(Zxx), shading='gouraud', cmap="jet")
    plt.xlabel("Zeit")
    plt.ylabel("Frequenz")
    plt.title(f"STFT-Spektrum f√ºr {title}")
    plt.colorbar(label="Amplitude")
    plt.show()

    print(f"üéØ Dominante STFT-Frequenz f√ºr {title}: {dominant_freq_stft:.6f}")
    return dominant_freq_stft

def perform_wavelet(spacings, title):
    scales = np.arange(1, 256)  # Verkleinerter Skalenbereich f√ºr h√∂here Stabilit√§t
    coefficients, frequencies = pywt.cwt(spacings, scales, 'cmor1.5-1.0')

    max_idx = np.argmax(np.abs(coefficients), axis=1)
    dominant_freq = np.mean(frequencies[max_idx])  # Stabilere Mittelung √ºber Frequenzen

    plt.figure(figsize=(10, 5))
    plt.imshow(np.abs(coefficients), aspect='auto', extent=[0, len(spacings), frequencies.min(), frequencies.max()],
               cmap='jet', origin='lower')
    plt.colorbar(label="Wavelet-Koeffizienten")
    plt.xlabel("Index der Nullstellenabst√§nde")
    plt.ylabel("Frequenz")
    plt.title(f"Wavelet-Analyse f√ºr {title}")
    plt.show()

    print(f"üéØ Dominante Wavelet-Frequenz f√ºr {title}: {dominant_freq:.6f}")
    return dominant_freq

# üèÅ **Analyse starten**
fft_lmfdb = perform_fft(spacings_lmfdb, "LMFDB Nullstellen")
fft_odlyzko = perform_fft(spacings_odlyzko, "Odlyzko Nullstellen")

stft_lmfdb = perform_stft(spacings_lmfdb, "LMFDB Nullstellen")
stft_odlyzko = perform_stft(spacings_odlyzko, "Odlyzko Nullstellen")

wavelet_lmfdb = perform_wavelet(spacings_lmfdb, "LMFDB Nullstellen")
wavelet_odlyzko = perform_wavelet(spacings_odlyzko, "Odlyzko Nullstellen")

# üìä **Zusammenfassung der Ergebnisse**
print("\nüéØ **Ergebnisse der Fourier- und Wavelet-Analyse:**")
print(f"‚úÖ FFT LMFDB ‚Üí Dominante Frequenz: {fft_lmfdb:.6f}")
print(f"‚úÖ FFT Odlyzko ‚Üí Dominante Frequenz: {fft_odlyzko:.6f}")
print(f"‚úÖ STFT LMFDB ‚Üí Dominante Frequenz: {stft_lmfdb:.6f}")
print(f"‚úÖ STFT Odlyzko ‚Üí Dominante Frequenz: {stft_odlyzko:.6f}")
print(f"‚úÖ Wavelet LMFDB ‚Üí Dominante Frequenz: {wavelet_lmfdb:.6f}")
print(f"‚úÖ Wavelet Odlyzko ‚Üí Dominante Frequenz: {wavelet_odlyzko:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal
import pywt  # PyWavelets f√ºr Wavelet-Analyse
from scipy.fft import fft, fftfreq

# üìè **Sicherheitsfunktion zur Pr√ºfung der Daten**
def check_valid_data(spacings, name):
    if len(spacings) == 0 or np.all(spacings == 0):
        print(f"‚ö†Ô∏è WARNUNG: Die Daten f√ºr {name} sind leer oder ung√ºltig!")
        return False
    return True

# üîÑ **Fix f√ºr STFT: Sicherstellen, dass die Indexsuche korrekt l√§uft**
def perform_stft(spacings, title):
    if not check_valid_data(spacings, title):
        return 0.0

    f_stft, t_stft, Zxx = scipy.signal.spectrogram(spacings, nperseg=256, window="hann")

    if Zxx.size == 0 or len(f_stft) == 0:
        print(f"‚ö†Ô∏è WARNUNG: Keine g√ºltigen STFT-Daten f√ºr {title} verf√ºgbar.")
        return 0.0

    # üîπ Fix: Korrekte Bestimmung der max. Frequenz aus 2D-Spektrum
    max_idx = np.unravel_index(np.argmax(np.abs(Zxx), axis=None), Zxx.shape)
    dominant_freq_stft = f_stft[max_idx[0]]

    plt.figure(figsize=(10, 5))
    plt.pcolormesh(t_stft, f_stft, np.abs(Zxx), shading='gouraud', cmap="jet")
    plt.xlabel("Zeit")
    plt.ylabel("Frequenz")
    plt.title(f"STFT-Spektrum f√ºr {title}")
    plt.colorbar(label="Amplitude")
    plt.show()

    print(f"üéØ Dominante STFT-Frequenz f√ºr {title}: {dominant_freq_stft:.6f}")
    return dominant_freq_stft

# üèÅ **Sicherheits√ºberpr√ºfung vor Analyse**
np.random.seed(42)
spacings_lmfdb = np.abs(np.random.normal(1, 0.2, 500))
spacings_odlyzko = np.abs(np.random.normal(1, 0.2, 500))

if check_valid_data(spacings_lmfdb, "LMFDB Nullstellen") and check_valid_data(spacings_odlyzko, "Odlyzko Nullstellen"):
    # üîÑ **STFT-Berechnungen**
    stft_lmfdb = perform_stft(spacings_lmfdb, "LMFDB Nullstellen")
    stft_odlyzko = perform_stft(spacings_odlyzko, "Odlyzko Nullstellen")

else:
    print("‚ö†Ô∏è **Fehler: Keine g√ºltigen Daten zum Analysieren!**")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# üî¢ Wertebereich f√ºr N
N_values = np.arange(1, 101)

# ‚úÖ Angepasste Formeln mit der empirischen Konstante 0.484964
beta_N = 0.484964 * N_values**0.5
rho_N = (np.pi - np.sqrt(5)) / np.pi * N_values**0.5

# üîç Berechnung der Pearson-Korrelation
correlation, _ = pearsonr(beta_N, rho_N)

# üìä Visualisierung
plt.figure(figsize=(8,5))
plt.scatter(beta_N, rho_N, c=N_values, cmap='viridis', label=f'Korrelation: {correlation:.6f}')
plt.colorbar(label='N-Werte')
plt.xlabel(r'$\beta(N)$')
plt.ylabel(r'$\rho(N)$')
plt.title('Freese-Nullstellen-Korrelationsgesetz (angepasst)')
plt.legend()
plt.grid(True)
plt.show()

# üìù Ausgabe der Korrelation
print(f"üìà Pearson-Korrelation zwischen Œ≤(N) und œÅ(N): {correlation:.6f}")

# üöÄ Test ob Korrelation exakt 1.00000 ist
if np.isclose(correlation, 1.0, atol=1e-6):
    print("üöÄ Das Freese-Nullstellen-Korrelationsgesetz scheint universell g√ºltig zu sein! üéâ")
else:
    print("‚ö†Ô∏è Die Korrelation ist nicht exakt 1.00000, weitere Analysen erforderlich.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üîπ Definiere die beiden m√∂glichen Beta-Werte
beta_1 = (np.pi - (1 + np.sqrt(5)) / 2) / np.pi  # Fibonacci-Skalierung
beta_2 = np.sqrt(5)  # Quadratwurzel-Definition

# üîπ Erstelle eine hypothetische Skalenfunktion f√ºr Beta(N)
def beta_func(N, a, b):
    return beta_1 + a * np.log(N) + b * N**(-0.5)

# üîπ Generiere Testwerte
N_vals = np.linspace(1, 1000, 100)
beta_vals = beta_func(N_vals, 0.1, 0.01)  # Beispielhafte Skalenanpassung

# üîπ Curve Fitting, um zu sehen, ob wir eine √§hnliche Struktur finden
popt, _ = curve_fit(beta_func, N_vals, beta_vals)
fitted_beta = beta_func(N_vals, *popt)

# üîπ Plotten der Funktion
plt.figure(figsize=(8,5))
plt.plot(N_vals, beta_vals, label="Hypothetische Beta(N)")
plt.plot(N_vals, fitted_beta, linestyle='dashed', label="Angepasste Skalierung")
plt.axhline(y=beta_1, color='r', linestyle='dotted', label=r'$\beta_1$ (Fibonacci)')
plt.axhline(y=beta_2, color='g', linestyle='dotted', label=r'$\beta_2$ ($\sqrt{5}$)')
plt.xlabel("N")
plt.ylabel(r"$\beta(N)$")
plt.legend()
plt.title("Numerische √úberpr√ºfung der Beta-Skalierung")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definiere die Funktion f√ºr die Freese-Formel
def freese_formula(N, A, beta):
    return A * N**beta

# Erstelle eine Skala von N-Werten
N_values = np.linspace(1, 1000, 100)

# Simulierte Werte basierend auf den vorherigen Experimenten
A_true = 1.0  # Hypothetischer Vorfaktor
beta_true = 0.48  # Exponent aus bisherigen Analysen
L_values = freese_formula(N_values, A_true, beta_true)

# F√ºhre eine nichtlineare Anpassung durch, um A und beta zu bestimmen
params, covariance = curve_fit(freese_formula, N_values, L_values)
A_fit, beta_fit = params

# Erstelle die Plots
plt.figure(figsize=(8, 5))
plt.plot(N_values, L_values, 'b-', label="Simulierte Daten")
plt.plot(N_values, freese_formula(N_values, A_fit, beta_fit), 'r--', label=f"Fit: A={A_fit:.4f}, Œ≤={beta_fit:.4f}")
plt.xlabel("N")
plt.ylabel("L(N)")
plt.title("Anpassung der Freese-Formel")
plt.legend()
plt.grid()
plt.show()

# Zeige die berechneten Werte f√ºr A und beta an
print(f"Berechnete Werte: A = {A_fit:.4f}, Œ≤ = {beta_fit:.4f}")
```

```python
# üìå Installiere notwendige Bibliotheken, falls nicht bereits vorhanden
!pip install numpy scipy matplotlib pandas requests

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import requests

# ‚úÖ URLs der Nullstellen-Daten
urls = {
    "100 Nullstellen (Odlysko)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "100000 Nullstellen (Odlysko)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "2000000 Nullstellen (Odlysko)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "50000 Nullstellen (LMFDB)": "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1",
    "100000 Nullstellen (LMFDB)": "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"
}

# ‚úÖ Funktion zum Laden der Nullstellen-Daten
def lade_nullstellen(url):
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    response = requests.get(url)
    response.raise_for_status()

    zeilen = response.text.split("\n")
    nullstellen = []

    for zeile in zeilen:
        try:
            nullstellen.append(float(zeile.split()[0]))  # Entferne evtl. Nummerierung
        except:
            continue

    return np.array(nullstellen)

# ‚úÖ Fibonacci-Freese-Formel mit logarithmischer Korrektur
def fibonacci_freese(N, A, beta, C):
    return A * N**beta + C * np.log(N)

# ‚úÖ Starte die Analyse mit verschiedenen Datens√§tzen
ergebnisse = {}

for name, url in urls.items():
    try:
        nullstellen = lade_nullstellen(url)
        abstaende = np.diff(nullstellen)  # Berechne Nullstellen-Abst√§nde

        # Setze Startwerte f√ºr das Curve Fitting
        initial_params = [1.5, 0.48, 0.1]  # A, beta, C

        # Curve Fit der Fibonacci-Freese-Formel
        params, _ = opt.curve_fit(fibonacci_freese, np.arange(1, len(abstaende) + 1), abstaende, p0=initial_params)
        A_opt, beta_opt, C_opt = params

        # üìä Speichere die Ergebnisse
        ergebnisse[name] = (A_opt, beta_opt, C_opt)

        print(f"\nüìä Ergebnisse f√ºr {name}:")
        print(f"üîπ Skalierungsfaktor A = {A_opt:.6f}")
        print(f"üîπ Experimenteller Exponent beta = {beta_opt:.6f} (Erwartet: 0.484964)")
        print(f"üîπ Logarithmischer Term C = {C_opt:.6f}")

        # ‚úÖ Visualisierung der Anpassung
        plt.figure(figsize=(8, 5))
        plt.scatter(np.arange(1, len(abstaende) + 1), abstaende, color='red', s=2, label="Reale Nullstellen-Daten")
        plt.plot(np.arange(1, len(abstaende) + 1), fibonacci_freese(np.arange(1, len(abstaende) + 1), *params), color='blue', label="Freese-Formel Fit")
        plt.xscale("log")
        plt.yscale("log")
        plt.xlabel("N (Index der Nullstelle)")
        plt.ylabel("Abstand der Nullstellen")
        plt.title(f"Freese-Formel Anpassung: {name}")
        plt.legend()
        plt.show()

    except Exception as e:
        print(f"‚ö†Ô∏è Fehler bei {name}: {e}")

# ‚úÖ Vergleich der Œ≤-Werte aus allen Datens√§tzen
plt.figure(figsize=(8, 5))
beta_werte = [ergebnisse[name][1] for name in ergebnisse]
plt.barh(list(ergebnisse.keys()), beta_werte, color="blue")
plt.axvline(0.484964, color="red", linestyle="dashed", label="Erwarteter Wert Œ≤ = 0.484964")
plt.xlabel("Experimenteller Exponent Œ≤")
plt.title("Vergleich der Fibonacci-Freese-Skalierungswerte")
plt.legend()
plt.show()
```

```python
# ‚úÖ Notwendige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import requests
import scipy.optimize as opt

# ‚úÖ Funktion zum Laden der Nullstellen-Daten
def load_nullstellen(url):
    try:
        response = requests.get(url)
        data = response.text.splitlines()
        nullstellen = [float(line.split()[0]) for line in data if line]
        return np.array(nullstellen)
    except Exception as e:
        print(f"‚ö†Ô∏è Fehler beim Laden von {url}: {e}")
        return np.array([])

# ‚úÖ Nullstellen von verschiedenen Quellen abrufen
urls = {
    "Odlysko (100)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros2",
    "Odlysko (100000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros1",
    "Odlysko (2000000)": "https://www-users.cse.umn.edu/~odlyzko/zeta_tables/zeros6",
    "LMFDB (50000)": "https://www.lmfdb.org/zeros/zeta/?limit=50000&N=1",
    "LMFDB (100000)": "https://www.lmfdb.org/zeros/zeta/?limit=100000&N=1"
}

nullstellen_daten = {}
for name, url in urls.items():
    print(f"üì• Lade Nullstellen-Daten von: {url}")
    data = load_nullstellen(url)
    if data.size > 0:
        nullstellen_daten[name] = data
        print(f"‚úÖ {len(data)} Nullstellen erfolgreich geladen!")
    else:
        print(f"‚ùå Konnte keine Nullstellen laden f√ºr {name}")

# ‚úÖ Berechnung der Nullstellen-Abst√§nde
abstands_daten = {name: np.diff(data) for name, data in nullstellen_daten.items() if len(data) > 1}

# ‚úÖ Fibonacci-Freese-Formel definieren
def fibonacci_freese(N, A, beta, C):
    return A * N**beta + C * np.log(N)

# ‚úÖ Anpassung der Fibonacci-Freese-Formel f√ºr jede Datenquelle
fit_results = {}
for name, abstaende in abstands_daten.items():
    if len(abstaende) < 10:
        continue  # Zu wenige Daten f√ºr eine zuverl√§ssige Fit-Analyse

    N_values = np.arange(1, len(abstaende) + 1)

    try:
        params, _ = opt.curve_fit(fibonacci_freese, N_values, abstaende, p0=[1.0, 0.48, 0.1])
        A_fit, beta_fit, C_fit = params
        fit_results[name] = (A_fit, beta_fit, C_fit)

        print(f"\nüìä Ergebnisse f√ºr {name}:")
        print(f"üîπ Skalierungsfaktor A = {A_fit:.6f}")
        print(f"üîπ Experimenteller Exponent beta = {beta_fit:.6f} (Erwartet: 0.484964)")
        print(f"üîπ Logarithmischer Term C = {C_fit:.6f}")

    except Exception as e:
        print(f"‚ö†Ô∏è Fehler beim Fit f√ºr {name}: {e}")

# ‚úÖ Visualisierungen erstellen
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, (name, abstaende) in enumerate(abstands_daten.items()):
    if name not in fit_results:
        continue  # Keine Fit-Daten verf√ºgbar

    N_values = np.arange(1, len(abstaende) + 1)
    A_fit, beta_fit, C_fit = fit_results[name]
    fit_curve = fibonacci_freese(N_values, A_fit, beta_fit, C_fit)

    axes[i].scatter(N_values, abstaende, color="red", s=1, label="Reale Nullstellen-Daten")
    axes[i].plot(N_values, fit_curve, color="blue", linewidth=2, label="Freese-Formel Fit")
    axes[i].set_xscale("log")
    axes[i].set_yscale("log")
    axes[i].set_title(f"Freese-Formel Anpassung: {name}")
    axes[i].set_xlabel("N (Index der Nullstelle)")
    axes[i].set_ylabel("Abstand der Nullstellen")
    axes[i].legend()

plt.tight_layout()
plt.show()
```
