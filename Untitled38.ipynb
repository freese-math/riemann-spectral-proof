```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import factorint

# üîπ Liouville-Funktion definieren
def liouville(n):
    """Berechnet die Liouville-Funktion mit Primfaktorzerlegung."""
    factors = factorint(n)  # Primfaktorzerlegung von n
    exponent_sum = sum(factors.values())  # Summe der Exponenten
    return (-1) ** exponent_sum  # Liouville-Funktion

# üîπ Siegel-Theta-Funktion
def theta_siegel(t, max_n=100):
    """Berechnet die Siegel-Theta-Funktion f√ºr gegebenes t."""
    return sum(liouville(n) * np.exp(-n**2 * t) for n in range(1, max_n + 1))

# üîπ Werte berechnen
t_values = np.linspace(0.01, 2, 200)
theta_values = [theta_siegel(t) for t in t_values]

# üîπ Plot der Funktion
plt.figure(figsize=(8,5))
plt.plot(t_values, theta_values, label="Siegel-Theta-Funktion", color='b')
plt.xlabel("$t$")
plt.ylabel("$\Theta(t)$")
plt.title("Siegel-Theta-Funktion")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import factorint

# üîπ Liouville-Funktion f√ºr Siegel-Theta
def liouville(n):
    """Berechnet die Liouville-Funktion mit Primfaktorzerlegung."""
    factors = factorint(n)  # Primfaktorzerlegung von n
    exponent_sum = sum(factors.values())  # Summe der Exponenten
    return (-1) ** exponent_sum  # Liouville-Funktion

# üîπ Siegel-Theta-Funktion
def theta_siegel(t, max_n=100):
    """Berechnet die Siegel-Theta-Funktion f√ºr gegebenes t."""
    return sum(liouville(n) * np.exp(-n**2 * t) for n in range(1, max_n + 1))

# üîπ Freese-Funktion (Beispielhafte Implementierung)
def freese_function(x, A=0.479028, B=0.500291, C=0.022936):
    """Freese-Funktion als Approximation der Nullstellen der Zeta-Funktion."""
    return A * x**B + C

# üîπ Beta-Skalen-Operator (Beispielmodell)
def beta_scale_operator(x, beta=0.484906, scale_factor=1.2):
    """Beta-Skalen-Operator als hypothetische Skalierungsfunktion."""
    return scale_factor * np.exp(-beta * x**2)

# üîπ Wertebereich f√ºr die Plots
x_values = np.linspace(0.01, 10, 300)
theta_values = [theta_siegel(t) for t in x_values]
freese_values = freese_function(x_values)
beta_values = beta_scale_operator(x_values)

# üîπ Grafik erzeugen
plt.figure(figsize=(10,6))
plt.plot(x_values, theta_values, label="Siegel-Theta-Funktion", color='b')
plt.plot(x_values, freese_values, label="Freese-Funktion (FFF)", color='r', linestyle='--')
plt.plot(x_values, beta_values, label="Beta-Skalen-Operator", color='g', linestyle=':')

plt.xlabel("$x$")
plt.ylabel("Funktionwerte")
plt.title("Vergleich: Siegel-Theta, Freese-Funktion, Beta-Skalen-Operator")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

# üìå Lade die echten Nullstellen der Zeta-Funktion aus Datei
# (Angenommen, wir haben eine Liste von √ºber 2 Millionen Nullstellen)
file_path = "/content/drive/MyDrive/zeros6.txt"  # Pfad zur Datei mit Nullstellen
nullstellen = np.loadtxt(file_path)  # Nullstellen laden

# üîπ Berechne Abst√§nde zwischen aufeinanderfolgenden Nullstellen
abst√§nde = np.diff(nullstellen)

# üîπ Hypothese: Skalenrelationen zwischen Nullstellen mit Beta-Skalen-Operator
def beta_skala(n, A, B):
    """Hypothetische Beta-Skalen-Funktion zur Modellierung der Nullstellen-Abst√§nde"""
    return A * (n ** B)

# üîπ Fitten der Skalenstruktur mit der Beta-Skalen-Funktion
n_values = np.arange(1, len(abst√§nde) + 1)
popt, _ = curve_fit(beta_skala, n_values, abst√§nde)

A_opt, B_opt = popt
print(f"üîπ Optimierte Beta-Skala-Parameter: A = {A_opt:.6f}, B = {B_opt:.6f}")

# üìä **Visualisierung der Skalierung**
plt.figure(figsize=(10, 5))
plt.scatter(n_values, abst√§nde, s=1, alpha=0.5, label="Echte Nullstellen-Abst√§nde")
plt.plot(n_values, beta_skala(n_values, A_opt, B_opt), color='red', linewidth=2, label="Beta-Skalen-Operator (Fit)")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Index der Nullstelle")
plt.ylabel("Abstand zur n√§chsten Nullstelle")
plt.title("Analyse der Beta-Skala-Selbst√§hnlichkeit in Nullstellen")
plt.legend()
plt.grid(True)
plt.show()

# üîé **Pr√ºfung auf Invarianz der Skala**
abweichungen = abst√§nde / beta_skala(n_values, A_opt, B_opt) - 1
max_abw = np.max(np.abs(abweichungen))
mean_abw = np.mean(np.abs(abweichungen))

print(f"üìå Maximale Abweichung von Beta-Skalen-Hypothese: {max_abw:.10f}")
print(f"üìå Durchschnittliche Abweichung: {mean_abw:.10f}")

# üåÄ Falls Abweichungen klein sind ‚Üí Beta-Skala k√∂nnte ein strukturelles Prinzip sein!
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.optimize import curve_fit

# üîπ Parameter der Freese-Funktion
A, B, C, D, E, F = 1.0, 0.5, 0.2, 1.0, 0.05, 0.1

# üîπ Definition der Freese-Funktion (FF)
def freese_function(x, A, B, C, D, E, F):
    return A * np.sin(B * x) + C * np.log(D * x + 1) + E * np.exp(-F * x)

# üîπ Generiere x-Werte (analytischer Bereich)
x_values = np.linspace(0.01, 50, 1000)
ff_values = freese_function(x_values, A, B, C, D, E, F)

# üîπ Lade echte Nullstellen der Zeta-Funktion (Beispiel-Daten)
zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:1000]  # Anpassung notwendig
ff_zeros = freese_function(zeta_zeros, A, B, C, D, E, F)

# üîπ FFT-Analyse der FF
ff_fft = np.abs(fft(ff_values))
freqs = np.fft.fftfreq(len(ff_values), d=(x_values[1] - x_values[0]))

# üîπ Test der Freese-Identit√§t
identity_test = np.exp(1j * np.pi * ff_values) + 1

# üîπ Plots
plt.figure(figsize=(12, 5))

# üîπ Vergleich: FF vs. Nullstellen
plt.subplot(1, 2, 1)
plt.plot(x_values, ff_values, label="Freese-Funktion", linestyle="dashed")
plt.scatter(zeta_zeros, ff_zeros, color='red', marker='x', label="FF(Nullstellen)")
plt.xlabel("x-Werte")
plt.ylabel("FF(x)")
plt.title("Vergleich: Freese-Funktion vs. Zeta-Nullstellen")
plt.legend()

# üîπ FFT-Analyse der FF
plt.subplot(1, 2, 2)
plt.plot(freqs[:len(ff_fft)//2], ff_fft[:len(ff_fft)//2])
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Spektralanalyse der Freese-Funktion")

plt.tight_layout()
plt.show()

# üîπ Test der Freese-Identit√§t
identity_check = np.max(np.abs(identity_test))
print(f"üìå Max. Abweichung von Freese-Identit√§t: {identity_check:.10f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# üîπ Freese-Funktion (FF) mit freien Parametern
def freese_function(x, A, B, C, D, E, F):
    return A * np.sin(B * x) + C * np.cos(D * x) + E * np.tanh(F * x)

# üîπ Ziel: Minimierung der maximalen Abweichung von der Euler-Identit√§t
def loss_function(params):
    A, B, C, D, E, F = params
    x_values = np.linspace(0, 10, 500)
    ff_values = freese_function(x_values, A, B, C, D, E, F)
    identity_deviation = np.abs(np.exp(1j * np.pi * ff_values) + 1)
    return np.max(identity_deviation)  # Maximale Abweichung minimieren

# üîπ Startwerte f√ºr die Parameter
initial_params = [1.0, 1.0, 0.5, 0.5, 0.1, 0.1]

# üîπ Optimierung der Freese-Parameter
result = minimize(loss_function, initial_params, method='Nelder-Mead')
optimized_params = result.x
max_deviation = result.fun

# üîπ Ergebnisse ausgeben
print(f"‚úÖ Optimierte Parameter: A={optimized_params[0]:.6f}, B={optimized_params[1]:.6f}, "
      f"C={optimized_params[2]:.6f}, D={optimized_params[3]:.6f}, "
      f"E={optimized_params[4]:.6f}, F={optimized_params[5]:.6f}")
print(f"üìå Maximale Abweichung von der Freese-Identit√§t: {max_deviation:.10f}")

# üîπ Visualisierung der optimierten Freese-Funktion
x_plot = np.linspace(0, 10, 500)
y_plot = freese_function(x_plot, *optimized_params)

plt.figure(figsize=(8, 5))
plt.plot(x_plot, y_plot, label="Optimierte Freese-Funktion", color='b')
plt.axhline(0.5, linestyle='--', color='r', label="Erwarteter Fixpunkt (1/2)")
plt.xlabel("x")
plt.ylabel("Freese-Funktion")
plt.title("Optimierte Freese-Funktion in Bezug auf die Freese-Identit√§t")
plt.legend()
plt.grid()
plt.show()
```

```python
# ‚úÖ Notwendige Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ‚úÖ Funktion zur Freese-Identit√§t definieren (mit optimierten Parametern)
def freese_identity(n, A, B, C, D, E, F):
    return A * np.sin(B * n) + C * np.cos(D * n) + E * np.sin(F * n)

# ‚úÖ Nullstellen aus Datei laden (Pfad anpassen!)
file_path = "/content/drive/MyDrive/zeros6.txt"
nullstellen = np.loadtxt(file_path)

# ‚úÖ Neu optimierte Parameter (ersetze durch die neuesten Werte)
A_opt, B_opt, C_opt, D_opt, E_opt, F_opt = 1.027696, 1.005247, 0.489052, 0.512409, 0.101819, 0.098923

# ‚úÖ Berechnung der Freese-Identit√§tswerte mit den optimierten Parametern
freese_values = freese_identity(np.arange(len(nullstellen)), A_opt, B_opt, C_opt, D_opt, E_opt, F_opt)

# ‚úÖ Beta-Skalen-Operator definieren (basiert auf Freese-Identit√§t)
def beta_scale_operator(n, alpha, beta):
    return alpha * freese_identity(n, A_opt, B_opt, C_opt, D_opt, E_opt, F_opt) + beta

# ‚úÖ Curve Fit zur Optimierung des Beta-Skalen-Operators
initial_params = [1.0, 0.0]  # Startwerte f√ºr alpha und beta
popt_beta, _ = curve_fit(beta_scale_operator, np.arange(len(nullstellen)), nullstellen, p0=initial_params)

# ‚úÖ Optimierte Werte f√ºr den Beta-Skalen-Operator
alpha_opt, beta_opt = popt_beta
print(f"üîπ Optimierte Beta-Skalen-Operator Parameter: Œ± = {alpha_opt:.6f}, Œ≤ = {beta_opt:.6f}")

# ‚úÖ Berechnung der Beta-Skalen-Werte mit optimierten Parametern
beta_values = beta_scale_operator(np.arange(len(nullstellen)), alpha_opt, beta_opt)

# ‚úÖ Berechnung der Abweichungen
abweichungen = np.abs(nullstellen - beta_values)
max_abw = np.max(abweichungen)
avg_abw = np.mean(abweichungen)

# ‚úÖ Ergebnisse ausgeben
print(f"üìå Maximale Abweichung von Beta-Skalen-Hypothese: {max_abw:.10f}")
print(f"üìå Durchschnittliche Abweichung: {avg_abw:.10f}")

# ‚úÖ Visualisierung der Ergebnisse
plt.figure(figsize=(10, 5))
plt.plot(nullstellen, label="Echte Nullstellen", marker='o', linestyle='', alpha=0.6)
plt.plot(beta_values, label="Beta-Skalen-Operator (Fit)", linestyle='--', linewidth=1.5, color='red')
plt.xlabel("Index der Nullstellen")
plt.ylabel("Wert")
plt.title("Beta-Skalen-Operator vs. Nullstellen der Riemannschen Zeta-Funktion")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üìå Lade die Nullstellen- und Beta-Werte-Daten
nullstellen = np.loadtxt("/content/drive/MyDrive/zeros6.txt")
beta_data = pd.read_csv("/content/drive/MyDrive/beta_scale_operator.csv")

# üìå Sicherstellen, dass beide Arrays die gleiche L√§nge haben
min_length = min(len(nullstellen), len(beta_data))
nullstellen = nullstellen[:min_length]
beta_values = beta_data["Beta_Scale"].values[:min_length]

# ‚úÖ 1. Potenzgesetz Fit: f(n) = A * n^B
def power_law_fit(n, A, B):
    return A * n**B

# ‚úÖ 2. Exponentielles Wachstum: f(n) = A * exp(B * n)
def exponential_fit(n, A, B):
    return A * np.exp(B * n)

# ‚úÖ Curve Fit durchf√ºhren (mit stabilen Startwerten)
try:
    # Potenzgesetz
    power_params, _ = curve_fit(power_law_fit, nullstellen, beta_values, p0=[1e6, -0.1], maxfev=10000)
    A_power, B_power = power_params
    print(f"üìå Potenzgesetz-Fit: A = {A_power:.6f}, B = {B_power:.6f}")

    # Exponentieller Fit
    exp_params, _ = curve_fit(exponential_fit, nullstellen, beta_values, p0=[1e6, -0.0001], maxfev=10000)
    A_exp, B_exp = exp_params
    print(f"üìå Exponentieller Fit: A = {A_exp:.6f}, B = {B_exp:.6f}")
except RuntimeError as e:
    print(f"‚ùå Curve-Fit Fehler: {e}")

# ‚úÖ Visualisierung der Fits
plt.figure(figsize=(8,5))
plt.scatter(nullstellen, beta_values, label="Echte Beta-Werte", s=1, alpha=0.5)
plt.plot(nullstellen, power_law_fit(nullstellen, A_power, B_power), 'r--', label=f"Power-Fit: A={A_power:.3f}, B={B_power:.3f}")
plt.plot(nullstellen, exponential_fit(nullstellen, A_exp, B_exp), 'g--', label=f"Exp-Fit: A={A_exp:.3f}, B={B_exp:.3f}")
plt.xlabel("Index n")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Echte Beta-Werte vs. Potenz- und Exponential-Fits")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üîπ Lade die Nullstellen der Riemannschen Zeta-Funktion
zeros_path = "/content/drive/MyDrive/zeros6.txt"
nullstellen = np.loadtxt(zeros_path)

# üîπ Lade die Beta-Skalen-Werte
beta_path = "/content/drive/MyDrive/beta_scale_operator.csv"
beta_data = pd.read_csv(beta_path)

# ‚úÖ Sicherstellen, dass die richtigen Spaltennamen verwendet werden
print("üìå Spalten in der CSV:", beta_data.columns)

# üõ† Falls die Spalten nicht korrekt erkannt werden, versuche eine automatische Erkennung
if "Beta_Scale" not in beta_data.columns:
    beta_column = [col for col in beta_data.columns if "Beta" in col][0]
else:
    beta_column = "Beta_Scale"

beta_values = beta_data[beta_column].values

# üîπ Skalierungspr√ºfung: L√§nge der Daten
print(f"üìå L√§nge Nullstellen: {len(nullstellen)}")
print(f"üìå L√§nge Beta-Werte: {len(beta_values)}")

# üîπ Berechnung der Abweichungen
abweichungen = np.abs(nullstellen - beta_values[:len(nullstellen)])

# üìä **Histogramm der Beta-Werte vs. Zeta-Nullstellen**
plt.figure(figsize=(8,5))
plt.hist(beta_values, bins=100, alpha=0.6, color='blue', label="Beta-Werte")
plt.hist(nullstellen, bins=100, alpha=0.6, color='orange', label="Zeta-Nullstellen")
plt.legend()
plt.xlabel("Werte")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm: Nullstellen vs. Beta-Werte")
plt.show()

# üìå **Berechnung der Fits** üìå
def power_law(n, A, B):
    return A * n ** B

def exponential_fit(n, A, B):
    return A * np.exp(B * n)

# ‚úÖ Power-Law Fit
params_power, _ = curve_fit(power_law, nullstellen, beta_values[:len(nullstellen)])
A_power, B_power = params_power
print(f"üìå Potenzgesetz-Fit: A = {A_power:.6f}, B = {B_power:.6f}")

# ‚úÖ Exponentieller Fit
params_exp, _ = curve_fit(exponential_fit, nullstellen, beta_values[:len(nullstellen)])
A_exp, B_exp = params_exp
print(f"üìå Exponentieller Fit: A = {A_exp:.6f}, B = {B_exp:.6f}")

# üìä **Vergleich der Fits mit echten Beta-Werten**
plt.figure(figsize=(8,5))
plt.scatter(nullstellen, beta_values[:len(nullstellen)], label="Echte Beta-Werte", s=1, alpha=0.5)
plt.plot(nullstellen, power_law(nullstellen, A_power, B_power), 'r--', label=f"Power-Fit: A={A_power:.3f}, B={B_power:.3f}")
plt.plot(nullstellen, exponential_fit(nullstellen, A_exp, B_exp), 'g--', label=f"Exp-Fit: A={A_exp:.3f}, B={B_exp:.3f}")
plt.xlabel("Index n")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Echte Beta-Werte vs. Fits")
plt.show()

# ‚úÖ **Hybrid-Fit (Kombination aus Power- und Exp-Fit)**
def hybrid_fit(n, A, B, C, D):
    return A * n ** B + C * np.exp(D * n)

params_hybrid, _ = curve_fit(hybrid_fit, nullstellen, beta_values[:len(nullstellen)], maxfev=10000)
A_hybrid, B_hybrid, C_hybrid, D_hybrid = params_hybrid
print(f"üìå Hybrid-Fit erfolgreich: A = {A_hybrid:.6f}, B = {B_hybrid:.6f}, C = {C_hybrid:.6f}, D = {D_hybrid:.6f}")

# üìä **Plot des Hybrid-Fits**
plt.figure(figsize=(8,5))
plt.scatter(nullstellen[:75000], beta_values[:75000], label="Echte Beta-Werte", s=1, alpha=0.5)
plt.plot(nullstellen[:75000], hybrid_fit(nullstellen[:75000], A_hybrid, B_hybrid, C_hybrid, D_hybrid), 'r--', label="Hybrid-Fit")
plt.xlabel("Index n")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Echte Beta-Werte vs. Hybrid-Fit")
plt.show()

# ‚úÖ **Finale Skalierungsanalyse**
skalen_faktor = np.mean(beta_values[:len(nullstellen)] / nullstellen)
print(f"üìå Gesch√§tzter Skalierungsfaktor: {skalen_faktor:.6f}")

plt.figure(figsize=(8,5))
plt.hist(beta_values[:len(nullstellen)] / skalen_faktor, bins=100, alpha=0.6, color='blue', label="Korrigierte Beta-Werte")
plt.hist(nullstellen, bins=100, alpha=0.6, color='orange', label="Zeta-Nullstellen")
plt.legend()
plt.xlabel("Werte")
plt.ylabel("H√§ufigkeit")
plt.title("Histogramm nach Skalierungskorrektur")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg

# üîπ Parameter f√ºr Matrixgr√∂√üe (z. B. 100x100)
N = 100

# üîπ Hermitesche Matrix mit Zufallselementen (Random Matrix Ansatz)
np.random.seed(42)  # Reproduzierbare Zufallszahlen
X = np.random.randn(N, N)
H = (X + X.T) / 2  # Symmetrisieren f√ºr Hermitesch-Eigenschaft

# üîπ Eigenwerte berechnen
eigenvalues = np.linalg.eigvalsh(H)  # Sortierte Eigenwerte

# üîπ Vergleich mit Zeta-Nullstellen
# (Hier nur als Beispiel eine lineare Skalierung f√ºr Testzwecke)
zeta_nullstellen = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:N]
zeta_scaled = zeta_nullstellen / np.max(zeta_nullstellen) * np.max(eigenvalues)

# üîπ Plot der Eigenwerte vs. Zeta-Nullstellen
plt.figure(figsize=(8,5))
plt.plot(range(N), eigenvalues, 'ro-', label="H-Eigenwerte")
plt.plot(range(N), zeta_scaled, 'bo-', label="Zeta-Nullstellen (skaliert)")
plt.xlabel("Index")
plt.ylabel("Eigenwert")
plt.legend()
plt.title("Vergleich: H-Eigenwerte vs. Zeta-Nullstellen")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

# ‚úÖ Beispiel-Daten (Ersetzen durch echte Zeta-Nullstellen oder Beta-Skalenwerte)
t_values = np.linspace(1, 100, 50)  # Indizes oder Nullstellen
h_values = np.log(t_values + 1e-8)  # üîπ Logarithmische Transformation (Vermeidung von log(0))

# ‚úÖ Normalisierung f√ºr Stabilit√§t
t_min, t_max = np.min(t_values), np.max(t_values)
h_min, h_max = np.min(h_values), np.max(h_values)

t_scaled = (t_values - t_min) / (t_max - t_min)  # Skalierung in [0,1]
h_scaled = (h_values - h_min) / (h_max - h_min)

# ‚úÖ Verschiedene Fit-Funktionen definieren
def log_fit(params, t, h_real):
    """ Logarithmischer Fit: h(t) = A + B log(t) """
    A, B = params
    h_pred = A + B * np.log(t + 1e-8)
    return h_pred - h_real

def power_fit(params, t, h_real):
    """ Potenzgesetz-Fit: h(t) = A * t^B """
    A, B = params
    h_pred = A * (t + 1e-8) ** B
    return h_pred - h_real

def sigmoid_fit(params, t, h_real):
    """ Sigmoid-Fit: h(t) = A / (1 + exp(-B (t - C))) """
    A, B, C = params
    h_pred = A / (1 + np.exp(-B * (t - C)))
    return h_pred - h_real

# ‚úÖ Adaptive Startwerte f√ºr bessere Konvergenz
A_start = np.median(h_scaled)
B_start = np.ptp(h_scaled) / np.ptp(np.log(t_scaled + 1e-8))
params_init_log = [A_start, B_start]
params_init_power = [A_start, B_start]
params_init_sigmoid = [1, 10, 0.5]  # Standardwerte f√ºr sigmoide Kurven

# ‚úÖ Least Squares Optimierung f√ºr verschiedene Modelle
result_log = least_squares(log_fit, params_init_log, args=(t_scaled, h_scaled), max_nfev=15000)
result_power = least_squares(power_fit, params_init_power, args=(t_scaled, h_scaled), max_nfev=15000)
result_sigmoid = least_squares(sigmoid_fit, params_init_sigmoid, args=(t_scaled, h_scaled), max_nfev=15000)

# ‚úÖ Optimierte Parameter extrahieren
A_log, B_log = result_log.x
A_power, B_power = result_power.x
A_sig, B_sig, C_sig = result_sigmoid.x

# ‚úÖ Berechnung der Fit-Werte
h_pred_log = A_log + B_log * np.log(t_scaled + 1e-8)
h_pred_power = A_power * (t_scaled + 1e-8) ** B_power
h_pred_sigmoid = A_sig / (1 + np.exp(-B_sig * (t_scaled - C_sig)))

# ‚úÖ Berechnung der Fit-Qualit√§t (R¬≤ und RMSE)
def calculate_fit_quality(h_real, h_pred):
    residuals = h_real - h_pred
    r_squared = 1 - (np.sum(residuals**2) / np.sum((h_real - np.mean(h_real))**2))
    rmse = np.sqrt(np.mean(residuals**2))
    return r_squared, rmse

r2_log, rmse_log = calculate_fit_quality(h_scaled, h_pred_log)
r2_power, rmse_power = calculate_fit_quality(h_scaled, h_pred_power)
r2_sigmoid, rmse_sigmoid = calculate_fit_quality(h_scaled, h_pred_sigmoid)

# ‚úÖ Ausgabe der Ergebnisse
print(f"üîπ Log-Fit: A = {A_log:.6f}, B = {B_log:.6f} | R¬≤ = {r2_log:.6f}, RMSE = {rmse_log:.6f}")
print(f"üîπ Power-Fit: A = {A_power:.6f}, B = {B_power:.6f} | R¬≤ = {r2_power:.6f}, RMSE = {rmse_power:.6f}")
print(f"üîπ Sigmoid-Fit: A = {A_sig:.6f}, B = {B_sig:.6f}, C = {C_sig:.6f} | R¬≤ = {r2_sigmoid:.6f}, RMSE = {rmse_sigmoid:.6f}")

# ‚úÖ Visualisierung der Anpassungen
plt.figure(figsize=(8,5))
plt.scatter(t_scaled, h_scaled, label="Echte Werte", color='blue', alpha=0.5)
plt.plot(t_scaled, h_pred_log, 'r--', label=f"Log-Fit (R¬≤={r2_log:.4f})")
plt.plot(t_scaled, h_pred_power, 'g--', label=f"Power-Fit (R¬≤={r2_power:.4f})")
plt.plot(t_scaled, h_pred_sigmoid, 'm--', label=f"Sigmoid-Fit (R¬≤={r2_sigmoid:.4f})")
plt.xlabel("t (normalisiert)")
plt.ylabel("h_eigen (normalisiert)")
plt.legend()
plt.title("Vergleich verschiedener Fit-Modelle")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# üìå Lade die Nullstellen- und Beta-Werte-Daten
nullstellen = np.loadtxt("/content/drive/MyDrive/zeros6.txt")
beta_data = pd.read_csv("/content/drive/MyDrive/beta_scale_operator.csv")

# üìå Sicherstellen, dass beide Arrays die gleiche L√§nge haben
min_length = min(len(nullstellen), len(beta_data))
nullstellen = nullstellen[:min_length]
beta_values = beta_data["Beta_Scale"].values[:min_length]

# ‚úÖ 1. Potenzgesetz Fit: f(n) = A * n^B
def power_law_fit(n, A, B):
    return A * n**B

# ‚úÖ 2. Exponentielles Wachstum: f(n) = A * exp(B * n)
def exponential_fit(n, A, B):
    return A * np.exp(B * n)

# ‚úÖ Curve Fit durchf√ºhren (mit stabilen Startwerten)
try:
    # Potenzgesetz
    power_params, _ = curve_fit(power_law_fit, nullstellen, beta_values, p0=[1e6, -0.1], maxfev=10000)
    A_power, B_power = power_params
    print(f"üìå Potenzgesetz-Fit: A = {A_power:.6f}, B = {B_power:.6f}")

    # Exponentieller Fit
    exp_params, _ = curve_fit(exponential_fit, nullstellen, beta_values, p0=[1e6, -0.0001], maxfev=10000)
    A_exp, B_exp = exp_params
    print(f"üìå Exponentieller Fit: A = {A_exp:.6f}, B = {B_exp:.6f}")
except RuntimeError as e:
    print(f"‚ùå Curve-Fit Fehler: {e}")

# ‚úÖ Visualisierung der Fits
plt.figure(figsize=(8,5))
plt.scatter(nullstellen, beta_values, label="Echte Beta-Werte", s=1, alpha=0.5)
plt.plot(nullstellen, power_law_fit(nullstellen, A_power, B_power), 'r--', label=f"Power-Fit: A={A_power:.3f}, B={B_power:.3f}")
plt.plot(nullstellen, exponential_fit(nullstellen, A_exp, B_exp), 'g--', label=f"Exp-Fit: A={A_exp:.3f}, B={B_exp:.3f}")
plt.xlabel("Index n")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Echte Beta-Werte vs. Potenz- und Exponential-Fits")
plt.show()
```
