```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import factorint

# 🔹 Liouville-Funktion definieren
def liouville(n):
    """Berechnet die Liouville-Funktion mit Primfaktorzerlegung."""
    factors = factorint(n)  # Primfaktorzerlegung von n
    exponent_sum = sum(factors.values())  # Summe der Exponenten
    return (-1) ** exponent_sum  # Liouville-Funktion

# 🔹 Siegel-Theta-Funktion
def theta_siegel(t, max_n=100):
    """Berechnet die Siegel-Theta-Funktion für gegebenes t."""
    return sum(liouville(n) * np.exp(-n**2 * t) for n in range(1, max_n + 1))

# 🔹 Werte berechnen
t_values = np.linspace(0.01, 2, 200)
theta_values = [theta_siegel(t) for t in t_values]

# 🔹 Plot der Funktion
plt.figure(figsize=(8,5))
plt.plot(t_values, theta_values, label="Siegel-Theta-Funktion", color='b')
plt.xlabel("$t$")
plt.ylabel("$\Theta(t)$")
plt.title("Siegel-Theta-Funktion")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import factorint

# 🔹 Liouville-Funktion für Siegel-Theta
def liouville(n):
    """Berechnet die Liouville-Funktion mit Primfaktorzerlegung."""
    factors = factorint(n)  # Primfaktorzerlegung von n
    exponent_sum = sum(factors.values())  # Summe der Exponenten
    return (-1) ** exponent_sum  # Liouville-Funktion

# 🔹 Siegel-Theta-Funktion
def theta_siegel(t, max_n=100):
    """Berechnet die Siegel-Theta-Funktion für gegebenes t."""
    return sum(liouville(n) * np.exp(-n**2 * t) for n in range(1, max_n + 1))

# 🔹 Freese-Funktion (Beispielhafte Implementierung)
def freese_function(x, A=0.479028, B=0.500291, C=0.022936):
    """Freese-Funktion als Approximation der Nullstellen der Zeta-Funktion."""
    return A * x**B + C

# 🔹 Beta-Skalen-Operator (Beispielmodell)
def beta_scale_operator(x, beta=0.484906, scale_factor=1.2):
    """Beta-Skalen-Operator als hypothetische Skalierungsfunktion."""
    return scale_factor * np.exp(-beta * x**2)

# 🔹 Wertebereich für die Plots
x_values = np.linspace(0.01, 10, 300)
theta_values = [theta_siegel(t) for t in x_values]
freese_values = freese_function(x_values)
beta_values = beta_scale_operator(x_values)

# 🔹 Grafik erzeugen
plt.figure(figsize=(10,6))
plt.plot(x_values, theta_values, label="Siegel-Theta-Funktion", color='b')
plt.plot(x_values, freese_values, label="Freese-Funktion (FFF)", color='r', linestyle='--')
plt.plot(x_values, beta_values, label="Beta-Skalen-Operator", color='g', linestyle=':')

plt.xlabel("$x$")
plt.ylabel("Funktionwerte")
plt.title("Vergleich: Siegel-Theta, Freese-Funktion, Beta-Skalen-Operator")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

# 📌 Lade die echten Nullstellen der Zeta-Funktion aus Datei
# (Angenommen, wir haben eine Liste von über 2 Millionen Nullstellen)
file_path = "/content/drive/MyDrive/zeros6.txt"  # Pfad zur Datei mit Nullstellen
nullstellen = np.loadtxt(file_path)  # Nullstellen laden

# 🔹 Berechne Abstände zwischen aufeinanderfolgenden Nullstellen
abstände = np.diff(nullstellen)

# 🔹 Hypothese: Skalenrelationen zwischen Nullstellen mit Beta-Skalen-Operator
def beta_skala(n, A, B):
    """Hypothetische Beta-Skalen-Funktion zur Modellierung der Nullstellen-Abstände"""
    return A * (n ** B)

# 🔹 Fitten der Skalenstruktur mit der Beta-Skalen-Funktion
n_values = np.arange(1, len(abstände) + 1)
popt, _ = curve_fit(beta_skala, n_values, abstände)

A_opt, B_opt = popt
print(f"🔹 Optimierte Beta-Skala-Parameter: A = {A_opt:.6f}, B = {B_opt:.6f}")

# 📊 **Visualisierung der Skalierung**
plt.figure(figsize=(10, 5))
plt.scatter(n_values, abstände, s=1, alpha=0.5, label="Echte Nullstellen-Abstände")
plt.plot(n_values, beta_skala(n_values, A_opt, B_opt), color='red', linewidth=2, label="Beta-Skalen-Operator (Fit)")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Index der Nullstelle")
plt.ylabel("Abstand zur nächsten Nullstelle")
plt.title("Analyse der Beta-Skala-Selbstähnlichkeit in Nullstellen")
plt.legend()
plt.grid(True)
plt.show()

# 🔎 **Prüfung auf Invarianz der Skala**
abweichungen = abstände / beta_skala(n_values, A_opt, B_opt) - 1
max_abw = np.max(np.abs(abweichungen))
mean_abw = np.mean(np.abs(abweichungen))

print(f"📌 Maximale Abweichung von Beta-Skalen-Hypothese: {max_abw:.10f}")
print(f"📌 Durchschnittliche Abweichung: {mean_abw:.10f}")

# 🌀 Falls Abweichungen klein sind → Beta-Skala könnte ein strukturelles Prinzip sein!
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.optimize import curve_fit

# 🔹 Parameter der Freese-Funktion
A, B, C, D, E, F = 1.0, 0.5, 0.2, 1.0, 0.05, 0.1

# 🔹 Definition der Freese-Funktion (FF)
def freese_function(x, A, B, C, D, E, F):
    return A * np.sin(B * x) + C * np.log(D * x + 1) + E * np.exp(-F * x)

# 🔹 Generiere x-Werte (analytischer Bereich)
x_values = np.linspace(0.01, 50, 1000)
ff_values = freese_function(x_values, A, B, C, D, E, F)

# 🔹 Lade echte Nullstellen der Zeta-Funktion (Beispiel-Daten)
zeta_zeros = np.loadtxt("/content/drive/MyDrive/zeros6.txt")[:1000]  # Anpassung notwendig
ff_zeros = freese_function(zeta_zeros, A, B, C, D, E, F)

# 🔹 FFT-Analyse der FF
ff_fft = np.abs(fft(ff_values))
freqs = np.fft.fftfreq(len(ff_values), d=(x_values[1] - x_values[0]))

# 🔹 Test der Freese-Identität
identity_test = np.exp(1j * np.pi * ff_values) + 1

# 🔹 Plots
plt.figure(figsize=(12, 5))

# 🔹 Vergleich: FF vs. Nullstellen
plt.subplot(1, 2, 1)
plt.plot(x_values, ff_values, label="Freese-Funktion", linestyle="dashed")
plt.scatter(zeta_zeros, ff_zeros, color='red', marker='x', label="FF(Nullstellen)")
plt.xlabel("x-Werte")
plt.ylabel("FF(x)")
plt.title("Vergleich: Freese-Funktion vs. Zeta-Nullstellen")
plt.legend()

# 🔹 FFT-Analyse der FF
plt.subplot(1, 2, 2)
plt.plot(freqs[:len(ff_fft)//2], ff_fft[:len(ff_fft)//2])
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Spektralanalyse der Freese-Funktion")

plt.tight_layout()
plt.show()

# 🔹 Test der Freese-Identität
identity_check = np.max(np.abs(identity_test))
print(f"📌 Max. Abweichung von Freese-Identität: {identity_check:.10f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# 🔹 Freese-Funktion (FF) mit freien Parametern
def freese_function(x, A, B, C, D, E, F):
    return A * np.sin(B * x) + C * np.cos(D * x) + E * np.tanh(F * x)

# 🔹 Ziel: Minimierung der maximalen Abweichung von der Euler-Identität
def loss_function(params):
    A, B, C, D, E, F = params
    x_values = np.linspace(0, 10, 500)
    ff_values = freese_function(x_values, A, B, C, D, E, F)
    identity_deviation = np.abs(np.exp(1j * np.pi * ff_values) + 1)
    return np.max(identity_deviation)  # Maximale Abweichung minimieren

# 🔹 Startwerte für die Parameter
initial_params = [1.0, 1.0, 0.5, 0.5, 0.1, 0.1]

# 🔹 Optimierung der Freese-Parameter
result = minimize(loss_function, initial_params, method='Nelder-Mead')
optimized_params = result.x
max_deviation = result.fun

# 🔹 Ergebnisse ausgeben
print(f"✅ Optimierte Parameter: A={optimized_params[0]:.6f}, B={optimized_params[1]:.6f}, "
      f"C={optimized_params[2]:.6f}, D={optimized_params[3]:.6f}, "
      f"E={optimized_params[4]:.6f}, F={optimized_params[5]:.6f}")
print(f"📌 Maximale Abweichung von der Freese-Identität: {max_deviation:.10f}")

# 🔹 Visualisierung der optimierten Freese-Funktion
x_plot = np.linspace(0, 10, 500)
y_plot = freese_function(x_plot, *optimized_params)

plt.figure(figsize=(8, 5))
plt.plot(x_plot, y_plot, label="Optimierte Freese-Funktion", color='b')
plt.axhline(0.5, linestyle='--', color='r', label="Erwarteter Fixpunkt (1/2)")
plt.xlabel("x")
plt.ylabel("Freese-Funktion")
plt.title("Optimierte Freese-Funktion in Bezug auf die Freese-Identität")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 📌 Lade die Nullstellen- und Beta-Werte-Daten
nullstellen = np.loadtxt("/content/drive/MyDrive/zeros6.txt")
beta_data = pd.read_csv("/content/drive/MyDrive/beta_scale_operator.csv")

# 📌 Sicherstellen, dass beide Arrays die gleiche Länge haben
min_length = min(len(nullstellen), len(beta_data))
nullstellen = nullstellen[:min_length]
beta_values = beta_data["Beta_Scale"].values[:min_length]

# ✅ 1. Potenzgesetz Fit: f(n) = A * n^B
def power_law_fit(n, A, B):
    return A * n**B

# ✅ 2. Exponentielles Wachstum: f(n) = A * exp(B * n)
def exponential_fit(n, A, B):
    return A * np.exp(B * n)

# ✅ Curve Fit durchführen (mit stabilen Startwerten)
try:
    # Potenzgesetz
    power_params, _ = curve_fit(power_law_fit, nullstellen, beta_values, p0=[1e6, -0.1], maxfev=10000)
    A_power, B_power = power_params
    print(f"📌 Potenzgesetz-Fit: A = {A_power:.6f}, B = {B_power:.6f}")

    # Exponentieller Fit
    exp_params, _ = curve_fit(exponential_fit, nullstellen, beta_values, p0=[1e6, -0.0001], maxfev=10000)
    A_exp, B_exp = exp_params
    print(f"📌 Exponentieller Fit: A = {A_exp:.6f}, B = {B_exp:.6f}")
except RuntimeError as e:
    print(f"❌ Curve-Fit Fehler: {e}")

# ✅ Visualisierung der Fits
plt.figure(figsize=(8,5))
plt.scatter(nullstellen, beta_values, label="Echte Beta-Werte", s=1, alpha=0.5)
plt.plot(nullstellen, power_law_fit(nullstellen, A_power, B_power), 'r--', label=f"Power-Fit: A={A_power:.3f}, B={B_power:.3f}")
plt.plot(nullstellen, exponential_fit(nullstellen, A_exp, B_exp), 'g--', label=f"Exp-Fit: A={A_exp:.3f}, B={B_exp:.3f}")
plt.xlabel("Index n")
plt.ylabel("Beta-Wert")
plt.legend()
plt.title("Vergleich: Echte Beta-Werte vs. Potenz- und Exponential-Fits")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

# ✅ Beispiel-Daten (Ersetzen durch echte Zeta-Nullstellen oder Beta-Skalenwerte)
t_values = np.linspace(1, 100, 50)  # Indizes oder Nullstellen
h_values = np.log(t_values + 1e-8)  # 🔹 Logarithmische Transformation (Vermeidung von log(0))

# ✅ Normalisierung für Stabilität
t_min, t_max = np.min(t_values), np.max(t_values)
h_min, h_max = np.min(h_values), np.max(h_values)

t_scaled = (t_values - t_min) / (t_max - t_min)  # Skalierung in [0,1]
h_scaled = (h_values - h_min) / (h_max - h_min)

# ✅ Verschiedene Fit-Funktionen definieren
def log_fit(params, t, h_real):
    """ Logarithmischer Fit: h(t) = A + B log(t) """
    A, B = params
    h_pred = A + B * np.log(t + 1e-8)
    return h_pred - h_real

def power_fit(params, t, h_real):
    """ Potenzgesetz-Fit: h(t) = A * t^B """
    A, B = params
    h_pred = A * (t + 1e-8) ** B
    return h_pred - h_real

def sigmoid_fit(params, t, h_real):
    """ Sigmoid-Fit: h(t) = A / (1 + exp(-B (t - C))) """
    A, B, C = params
    h_pred = A / (1 + np.exp(-B * (t - C)))
    return h_pred - h_real

# ✅ Adaptive Startwerte für bessere Konvergenz
A_start = np.median(h_scaled)
B_start = np.ptp(h_scaled) / np.ptp(np.log(t_scaled + 1e-8))
params_init_log = [A_start, B_start]
params_init_power = [A_start, B_start]
params_init_sigmoid = [1, 10, 0.5]  # Standardwerte für sigmoide Kurven

# ✅ Least Squares Optimierung für verschiedene Modelle
result_log = least_squares(log_fit, params_init_log, args=(t_scaled, h_scaled), max_nfev=15000)
result_power = least_squares(power_fit, params_init_power, args=(t_scaled, h_scaled), max_nfev=15000)
result_sigmoid = least_squares(sigmoid_fit, params_init_sigmoid, args=(t_scaled, h_scaled), max_nfev=15000)

# ✅ Optimierte Parameter extrahieren
A_log, B_log = result_log.x
A_power, B_power = result_power.x
A_sig, B_sig, C_sig = result_sigmoid.x

# ✅ Berechnung der Fit-Werte
h_pred_log = A_log + B_log * np.log(t_scaled + 1e-8)
h_pred_power = A_power * (t_scaled + 1e-8) ** B_power
h_pred_sigmoid = A_sig / (1 + np.exp(-B_sig * (t_scaled - C_sig)))

# ✅ Berechnung der Fit-Qualität (R² und RMSE)
def calculate_fit_quality(h_real, h_pred):
    residuals = h_real - h_pred
    r_squared = 1 - (np.sum(residuals**2) / np.sum((h_real - np.mean(h_real))**2))
    rmse = np.sqrt(np.mean(residuals**2))
    return r_squared, rmse

r2_log, rmse_log = calculate_fit_quality(h_scaled, h_pred_log)
r2_power, rmse_power = calculate_fit_quality(h_scaled, h_pred_power)
r2_sigmoid, rmse_sigmoid = calculate_fit_quality(h_scaled, h_pred_sigmoid)

# ✅ Ausgabe der Ergebnisse
print(f"🔹 Log-Fit: A = {A_log:.6f}, B = {B_log:.6f} | R² = {r2_log:.6f}, RMSE = {rmse_log:.6f}")
print(f"🔹 Power-Fit: A = {A_power:.6f}, B = {B_power:.6f} | R² = {r2_power:.6f}, RMSE = {rmse_power:.6f}")
print(f"🔹 Sigmoid-Fit: A = {A_sig:.6f}, B = {B_sig:.6f}, C = {C_sig:.6f} | R² = {r2_sigmoid:.6f}, RMSE = {rmse_sigmoid:.6f}")

# ✅ Visualisierung der Anpassungen
plt.figure(figsize=(8,5))
plt.scatter(t_scaled, h_scaled, label="Echte Werte", color='blue', alpha=0.5)
plt.plot(t_scaled, h_pred_log, 'r--', label=f"Log-Fit (R²={r2_log:.4f})")
plt.plot(t_scaled, h_pred_power, 'g--', label=f"Power-Fit (R²={r2_power:.4f})")
plt.plot(t_scaled, h_pred_sigmoid, 'm--', label=f"Sigmoid-Fit (R²={r2_sigmoid:.4f})")
plt.xlabel("t (normalisiert)")
plt.ylabel("h_eigen (normalisiert)")
plt.legend()
plt.title("Vergleich verschiedener Fit-Modelle")
plt.show()
```
