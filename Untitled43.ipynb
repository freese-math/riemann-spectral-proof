```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as sla
import psutil
import time

# ğŸ“Œ Zeta-Nullstellen laden
file_path = "/content/drive/MyDrive/zeros6.txt"

try:
    with open(file_path, "r") as f:
        zeta_zeros = np.array([float(line.strip()) for line in f.readlines() if line.strip()])[:1000000]
    print(f"âœ… Erfolgreich {len(zeta_zeros)} Zeta-Nullstellen geladen!")
except Exception as e:
    print(f"âŒ Fehler: {e}")
    zeta_zeros = np.linspace(0, 1000000, 1000000)

# ğŸ“Œ Speicherverbrauch vor Matrix-Erstellung
print(f"ğŸ“Œ Speicherverbrauch vor Matrix-Erstellung: {psutil.virtual_memory().used / 1e9:.2f} GB")

# ğŸ“Œ Sparse-Matrix erstellen
start_time = time.time()
N = len(zeta_zeros)
diagonals = [zeta_zeros, np.ones(N-1), np.ones(N-1)]
positions = [0, 1, -1]
H_sparse = sp.diags(diagonals, positions, format="csc")
print(f"âœ… Sparse-Hamilton-Matrix erstellt! â³ {time.time() - start_time:.2f} Sekunden")

# ğŸ“Œ Speicherverbrauch nach Matrix-Erstellung
print(f"ğŸ“Œ Speicherverbrauch nach Matrix-Erstellung: {psutil.virtual_memory().used / 1e9:.2f} GB")

# ğŸ“Œ Konditionszahl testen (nur fÃ¼r 100x100 Submatrix, da vollstÃ¤ndige Berechnung zu teuer)
submatrix = H_sparse[:100, :100].toarray()
cond_number = np.linalg.cond(submatrix)
print(f"ğŸ“Œ Konditionszahl (100x100 Submatrix): {cond_number:.2e}")

# ğŸ“Œ **Wichtiger Debug-Check**: Nur 10 Eigenwerte berechnen
try:
    print("â³ Berechne 10 Eigenwerte zum Test...")
    start_time = time.time()
    eigenvalues, _ = sla.eigsh(H_sparse, k=1, which="SM")
    print(f"âœ… Erste 10 Eigenwerte berechnet! â³ {time.time() - start_time:.2f} Sekunden")
except Exception as e:
    print(f"âŒ Fehler bei Eigenwertberechnung: {e}")
```

```python
# ğŸ“Œ Notwendige Bibliotheken installieren
!pip install numpy scipy matplotlib

# ğŸ“Œ Importiere Bibliotheken
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# ğŸ“Œ Zeta-Nullstellen aus Datei laden (Optimiert!)
file_path = "/content/drive/MyDrive/zeros6.txt"

try:
    with open(file_path, "r") as f:
        zeta_zeros = np.array([float(line.strip()) for line in f.readlines() if line.strip()])[:1000000]
    print(f"âœ… Erfolgreich {len(zeta_zeros)} echte Zeta-Nullstellen geladen!")
except Exception as e:
    print(f"âŒ Fehler beim Laden der Datei: {e}")
    zeta_zeros = np.linspace(0, 1000000, 1000000)  # Fallback-Daten

# ğŸ“Œ Hamilton-Operator definieren (Dichte Matrix)
N = len(zeta_zeros)
H = np.diag(zeta_zeros) + np.diag(np.ones(N-1), k=1) + np.diag(np.ones(N-1), k=-1)

print("âœ… Hamilton-Matrix erstellt!")

# ğŸ“Œ Eigenwerte berechnen (Stabile `eigh()`-Methode)
print("â³ Berechne Eigenwerte...")
eigenvalues, _ = la.eigh(H)
print("âœ… Eigenwerte erfolgreich berechnet!")

# ğŸ“Œ Selbstadjungiertheit Ã¼berprÃ¼fen
is_self_adjoint = np.allclose(H, H.T.conj(), atol=1e-10)
print(f"âœ… Selbstadjungiert: {is_self_adjoint}")

# ğŸ“Œ Erste 10 Eigenwerte anzeigen
print("ğŸ“Œ Erste 10 Eigenwerte des Hamilton-Operators:")
print(eigenvalues[:10])

# ğŸ“Œ Fourier-Analyse der Eigenwerte (Spektralanalyse)
fft_spectrum = np.fft.fft(eigenvalues)
frequencies = np.fft.fftfreq(len(eigenvalues))

# ğŸ“Œ Dominante Frequenzen extrahieren
dominant_indices = np.argsort(np.abs(fft_spectrum))[-5:]
dominant_frequencies = frequencies[dominant_indices]
dominant_amplitudes = np.abs(fft_spectrum[dominant_indices])

print("ğŸ”¥ Dominante Frequenzen der Eigenwerte:", np.round(dominant_frequencies, 6))
print("ğŸ”¥ Korrespondierende Amplituden:", np.round(dominant_amplitudes, 6))

# ğŸ“Œ Korrelation der Eigenwerte mit den Zeta-Nullstellen berechnen
correlation = np.corrcoef(eigenvalues, zeta_zeros)[0, 1]
print(f"âœ… Korrelation der Eigenwerte mit Zeta-Nullstellen: {correlation:.6f}")

# ğŸ“Œ Fehler in der Euler-Freese-IdentitÃ¤t berechnen
euler_freese_error = np.sum((eigenvalues - zeta_zeros) ** 2) / len(eigenvalues)
print(f"âœ… Fehler in der Euler-Freese-IdentitÃ¤t: {euler_freese_error:.6f}")

# ğŸ“Œ Plot: Eigenwerte vs. Index
plt.figure(figsize=(8, 5))
plt.plot(eigenvalues, "bo-", markersize=2, label="Eigenwerte")
plt.xlabel("Index")
plt.ylabel("Eigenwerte")
plt.title("Spektrum des Hamilton-Operators")
plt.legend()
plt.grid()
plt.show()

# ğŸ“Œ Plot: Fourier-Spektrum der Eigenwerte
plt.figure(figsize=(8, 5))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_spectrum)[:len(frequencies)//2], "r-")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale IntensitÃ¤t")
plt.title("Fourier-Spektrum der Eigenwerte")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# ğŸ”¹ 1ï¸âƒ£ Experimentelle Beta-Werte (Beispieldaten)
N = np.linspace(1, 1000, 1000)  # Index-Werte
beta_exp = 0.5 + 0.02 * np.sin(2 * np.pi * N / 100)  # Simulierte Beta-Oszillation

# ğŸ”¹ 2ï¸âƒ£ Verbessertes Planck-Spektrum
def planck_spectrum(N, T=5778):
    h = 6.62607015e-34  # Planck-Konstante (JouleÂ·s)
    c = 3e8  # Lichtgeschwindigkeit (m/s)
    k_B = 1.380649e-23  # Boltzmann-Konstante (J/K)

    Î» = 1e-6 * (N / max(N))  # WellenlÃ¤nge in Î¼m skaliert
    Î» = np.clip(Î», 1e-8, 1e-2)  # Untere Schranke setzen (z.B. 10 nm - 100 Î¼m)

    exponent = h*c / (Î» * k_B * T)
    exponent = np.clip(exponent, None, 100)  # Overflow verhindern
    return (2*h*c**2) / (Î»**5 * (np.exp(exponent) - 1))

planck_norm = planck_spectrum(N)
planck_norm /= np.max(planck_norm)  # Normierung

# ğŸ”¹ 3ï¸âƒ£ Fit-Funktion mit Planck, Feigenbaum und Feinstrukturkonstante
alpha = 1 / 137.035999  # Feinstrukturkonstante
delta_F = 4.669201  # Feigenbaum-Konstante

def beta_model(N, A, B, C, D):
    return A * planck_norm + B * alpha + C * delta_F + D

# ğŸ”¹ 4ï¸âƒ£ Fit durchfÃ¼hren
params, covariance = curve_fit(beta_model, N, beta_exp, p0=[0.1, 0.1, 0.1, 0.5])

# ğŸ”¹ 5ï¸âƒ£ Ergebnisse visualisieren
plt.figure(figsize=(8, 5))
plt.plot(N, beta_exp, 'b-', label='Experimentelle Î²-Werte')
plt.plot(N, beta_model(N, *params), 'r--', label='Fitted Î²-Modell')
plt.xlabel("Index N")
plt.ylabel("Î²(N)")
plt.legend()
plt.title("Strenger Fit von Beta-Werten mit Planck, Feigenbaum & Feinstrukturkonstante")
plt.grid()
plt.show()

# ğŸ”¹ 6ï¸âƒ£ Fit-Parameter ausgeben
print("Fit-Parameter (A, B, C, D):", params)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm
import statsmodels.api as sm

# ğŸ”¹ Beispiel: Experimentelle Beta-Werte
N = np.arange(1, 1001)
beta_exp = 0.5 + 0.02 * np.cos(2 * np.pi * N / 200)  # Simulierte Daten

# ğŸ”¹ Modellfunktion (angepasst an bisherige Fits)
def beta_model(N, A, B, C, D, E):
    return A + B * np.cos(C * N + D) + E * np.log(N)

# ğŸ”¹ Fit des Modells an die Daten
initial_params = [0.5, 0.02, 2*np.pi/200, 0, 0]
params_opt, _ = curve_fit(beta_model, N, beta_exp, p0=initial_params)

# ğŸ”¹ Modellierte Werte berechnen
beta_pred = beta_model(N, *params_opt)

# ğŸ”¹ Residuen berechnen
residuals = beta_exp - beta_pred

# ğŸ”¹ 1ï¸âƒ£ Streudiagramm der Residuen
plt.figure(figsize=(10, 4))
plt.plot(N, residuals, 'ro', markersize=2, alpha=0.6)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("Index N")
plt.ylabel("Residual")
plt.title("Streudiagramm der Residuen")
plt.grid()
plt.show()

# ğŸ”¹ 2ï¸âƒ£ Histogramm der Residuen (Normalverteilung prÃ¼fen)
plt.figure(figsize=(8, 4))
plt.hist(residuals, bins=30, density=True, alpha=0.6, color='b')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(residuals), np.std(residuals))
plt.plot(x, p, 'r', linewidth=2)
plt.xlabel("Residualwert")
plt.ylabel("Dichte")
plt.title("Histogramm der Residuen mit Normalverteilung")
plt.grid()
plt.show()

# ğŸ”¹ 3ï¸âƒ£ Autokorrelation der Residuen
sm.graphics.tsa.plot_acf(residuals, lags=50)
plt.title("Autokorrelation der Residuen")
plt.show()

# ğŸ”¹ 4ï¸âƒ£ Fourier-Analyse der Residuen (Fehlende Frequenzen?)
fft_residuals = np.fft.fft(residuals)
frequencies = np.fft.fftfreq(len(N))

plt.figure(figsize=(8, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_residuals)[:len(frequencies)//2], 'g')
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Residuen")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm
from statsmodels.tsa.stattools import acf

# Beispiel: Erzeugung von Beta-Werten (Falls keine echten Daten vorhanden sind)
N = np.arange(1, 1001)
def beta_function(n, A, B, C, D, E, F, G):
    return A * np.cos(B * n) + C * np.cos(D * n) + E * np.sin(F * n) + G

# Simulierte Beta-Werte mit Rauschen
true_params = [0.02, np.pi/100, 0.001, 2*np.pi/50, -3.5e-10, 9.92e-5, 5.51e-11]
beta_exp = beta_function(N, *true_params) + np.random.normal(scale=1e-10, size=len(N))

# Curve Fit
initial_params = [0.02, np.pi/100, 0.001, 2*np.pi/50, -3.5e-10, 9.92e-5, 5.51e-11]
params_opt, _ = curve_fit(beta_function, N, beta_exp, p0=initial_params)

# Berechnung der Residuen
residuals = beta_exp - beta_function(N, *params_opt)

# **1. Streudiagramm der Residuen**
plt.figure(figsize=(8,4))
plt.scatter(N, residuals, color='red', alpha=0.6, s=5)
plt.axhline(0, linestyle='--', color='black')
plt.title("Streudiagramm der Residuen")
plt.xlabel("Index N")
plt.ylabel("Residual")
plt.show()

# **2. Histogramm der Residuen mit Normalverteilung**
plt.figure(figsize=(8,4))
n, bins, _ = plt.hist(residuals, bins=30, density=True, color='blue', alpha=0.6)
mu, std = norm.fit(residuals)
pdf = norm.pdf(bins, mu, std)
plt.plot(bins, pdf, 'r-', label='Normalverteilung')
plt.title("Histogramm der Residuen mit Normalverteilung")
plt.xlabel("Residualwert")
plt.ylabel("Dichte")
plt.legend()
plt.show()

# **3. Autokorrelation der Residuen**
acf_values = acf(residuals, nlags=50, fft=True)
plt.figure(figsize=(6,4))
plt.stem(range(len(acf_values)), acf_values, use_line_collection=True)
plt.fill_between(range(len(acf_values)), -1.96/np.sqrt(len(N)), 1.96/np.sqrt(len(N)), color='blue', alpha=0.2)
plt.axhline(0, linestyle='-', color='black')
plt.title("Autokorrelation der Residuen")
plt.show()

# **4. Fourier-Analyse der Residuen**
fft_residuals = np.fft.fft(residuals)
frequencies = np.fft.fftfreq(len(residuals))
plt.figure(figsize=(8,4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_residuals)[:len(frequencies)//2], color='green')
plt.title("Fourier-Spektrum der Residuen")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.show()

# **Ergebnisse der optimierten Parameter**
print("Optimierte Parameter (A, B, C, D, E, F, G):", params_opt)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from statsmodels.tsa.stattools import acf

# Beispiel-Daten generieren
N = np.arange(1000)
true_beta = 0.5 + 0.02 * np.sin(0.03141593 * N)
noise = np.random.normal(0, 0.002, len(N))
beta_exp = true_beta + noise

# Fit-Modelle definieren
def simple_model(N, A, B, C):
    return A + B * np.sin(C * N)

def oscillating_model(N, A, B, C, D, E, F, G):
    return A + B * np.sin(C * N) + D * np.sin(E * N + F) + G * N

# Fit der Modelle
params_simple, _ = curve_fit(simple_model, N, beta_exp, p0=[0.5, 0.02, 0.03141593])
params_osc, _ = curve_fit(oscillating_model, N, beta_exp, p0=[0.5, 0.02, 0.03141593, 0, 0, 0, 0])

# Vorhersagen der Modelle
beta_simple_fit = simple_model(N, *params_simple)
beta_osc_fit = oscillating_model(N, *params_osc)

# Residuen berechnen
residuals_simple = beta_exp - beta_simple_fit
residuals_osc = beta_exp - beta_osc_fit

# Fehlerberechnung
mse_simple = np.mean(residuals_simple**2)
mse_osc = np.mean(residuals_osc**2)
std_simple = np.std(residuals_simple)
std_osc = np.std(residuals_osc)

# Autokorrelation der Residuen
acf_simple = acf(residuals_simple, nlags=50, fft=True)
acf_osc = acf(residuals_osc, nlags=50, fft=True)

# Fourier-Transformation der Residuen
freqs = np.fft.rfftfreq(len(N))
fft_simple = np.abs(np.fft.rfft(residuals_simple))
fft_osc = np.abs(np.fft.rfft(residuals_osc))

# Ergebnisse visualisieren
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# MSE Vergleich
axes[0, 0].bar(['Einfaches Modell', 'Oszillierendes Modell'], [mse_simple, mse_osc], color=['blue', 'red'])
axes[0, 0].set_title('Mittlerer quadratischer Fehler (MSE)')
axes[0, 0].set_ylabel('MSE-Wert')

# Standardabweichung Vergleich
axes[0, 1].bar(['Einfaches Modell', 'Oszillierendes Modell'], [std_simple, std_osc], color=['blue', 'red'])
axes[0, 1].set_title('Standardabweichung der Residuen')
axes[0, 1].set_ylabel('Standardabweichung')

# Autokorrelation
axes[1, 0].stem(range(len(acf_simple)), acf_simple, linefmt='b-', markerfmt='bo', basefmt='k-')
axes[1, 0].set_title('Autokorrelation der Residuen (Einfaches Modell)')
axes[1, 0].set_xlabel('Lag')
axes[1, 0].set_ylabel('Autokorrelation')

axes[1, 1].stem(range(len(acf_osc)), acf_osc, linefmt='r-', markerfmt='ro', basefmt='k-')
axes[1, 1].set_title('Autokorrelation der Residuen (Oszillierendes Modell)')
axes[1, 1].set_xlabel('Lag')
axes[1, 1].set_ylabel('Autokorrelation')

# Fourier-Analyse
axes[2, 0].plot(freqs, fft_simple, color='blue')
axes[2, 0].set_title('Fourier-Spektrum der Residuen (Einfaches Modell)')
axes[2, 0].set_xlabel('Frequenz')
axes[2, 0].set_ylabel('Amplitude')

axes[2, 1].plot(freqs, fft_osc, color='red')
axes[2, 1].set_title('Fourier-Spektrum der Residuen (Oszillierendes Modell)')
axes[2, 1].set_xlabel('Frequenz')
axes[2, 1].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Entscheidung basierend auf Fehlerwerten
best_model = "Einfaches Modell" if mse_simple < mse_osc else "Oszillierendes Modell"
print("Das beste Modell ist:", best_model)
```
