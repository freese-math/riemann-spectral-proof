```python
# ✅ Installation
!pip install mpmath sympy

# ✅ Imports
import numpy as np
import matplotlib.pyplot as plt
import torch
from mpmath import zeta, diff, li, re

# ✅ Gerät (GPU nutzen, wenn verfügbar)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("✅ Verwende Gerät:", device)

# ✅ Parameter
MAX_N = 100
x_vals = np.linspace(2, 100, 100)

# ✅ Lade rekonstruktive β-Skala
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta_vals = np.loadtxt(beta_path, comments="#", max_rows=MAX_N)
beta = torch.tensor(beta_vals, dtype=torch.float32, device=device)

# ✅ Frequenzen definieren (Odlyzko-Zeta-Nullstellen)
frequencies = torch.tensor([
    9.994743e-06, 9.495006e-06, 8.995268e-06, 8.495531e-06, 7.995794e-06,
    7.496057e-06, 6.996320e-06, 6.496583e-06, 5.996846e-06, 5.497109e-06,
    4.997371e-06, 4.497634e-06, 3.997897e-06, 3.498160e-06, 2.998423e-06,
    2.498686e-06, 1.998949e-06, 9.994743e-07, 1.499211e-06, 4.997371e-07
], device=device)

# ✅ Umrechnen in γ und ρ = 0.5 + iγ
gamma = 2 * np.pi * frequencies[:MAX_N].cpu().numpy()
rho = 0.5 + 1j * gamma

# ✅ Formel (76) Original (mit Zeta und Ableitung)
def L_beta_full(beta, gamma, x_vals):
    result = []
    for x in x_vals:
        x_log = np.log(x)
        summe = 0
        for k in range(len(beta)):
            rho_k = 0.5 + 1j * gamma[k]
            try:
                term = beta[k].item() * (x ** rho_k) * zeta(2 * rho_k) / (rho_k * diff(zeta, rho_k))
                summe += term.real
            except:
                continue
        result.append(1 + summe)
    return np.array(result)

# ✅ Vereinfachte Formel (nur Cosinus-Modell)
def L_beta_simplified(beta, gamma, x_vals):
    result = []
    for x in x_vals:
        x_log = np.log(x)
        cos_sum = torch.sum(beta * torch.cos(torch.tensor(gamma) * x_log))
        result.append((x**0.5) * cos_sum.item())
    return np.array(result)

# ✅ Klassische ψ(x)
psi_vals = np.array([float(sum([re(li(x**p)) for p in range(1, 21)])) for x in x_vals])

# ✅ Werte berechnen
L_full = L_beta_full(beta, gamma, x_vals)
L_simple = L_beta_simplified(beta, gamma, x_vals)

# ✅ Plot
plt.figure(figsize=(14, 6))
plt.plot(x_vals, psi_vals, label=r"$\psi(x)$ (klassisch)", color="black")
plt.plot(x_vals, L_full, label=r"$L(x)$ Formel (76) – vollständig", linestyle="--", color="blue")
plt.plot(x_vals, L_simple, label=r"$L(x)$ – vereinfacht mit $\cos(\gamma_k \log x)$", linestyle=":", color="orange")
plt.xlabel("x")
plt.ylabel("Wert")
plt.title("Vergleich: Liouville-Formel (vollständig vs. vereinfacht) gegen $\psi(x)$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ Installationen (falls nötig)
!pip install scipy numpy matplotlib

# ✅ Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, fftfreq

# ✅ Lorentz-Funktion definieren
def lorentzian(omega, A, omega_0, gamma):
    return A * gamma**2 / ((omega - omega_0)**2 + gamma**2)

# ✅ Mehrere Lorentzians summieren
def multi_lorentz(omega, *params):
    n_peaks = len(params) // 3
    y = np.zeros_like(omega)
    for i in range(n_peaks):
        A = params[3*i]
        omega_0 = params[3*i+1]
        gamma = params[3*i+2]
        y += lorentzian(omega, A, omega_0, gamma)
    return y

# ✅ Beispielsignal: Summe von drei Sinusschwingungen
N = 10000
T = 1.0
t = np.linspace(0, N*T, N)
signal = (
    3.0 * np.sin(2 * np.pi * 0.25615 * t) +
    2.0 * np.sin(2 * np.pi * 0.3097 * t) +
    1.5 * np.sin(2 * np.pi * 0.45092 * t)
)

# ✅ FFT berechnen
yf = np.abs(fft(signal))[:N//2]
xf = fftfreq(N, T)[0:N//2]

# ✅ Anfangsschätzung für drei Lorentz-Peaks
initial_guess = [
    max(yf), 0.25615, 0.01,   # Peak 1
    max(yf)/2, 0.3097, 0.01,  # Peak 2
    max(yf)/3, 0.4509, 0.01   # Peak 3
]

# ✅ Lorentz-Fit durchführen
popt, _ = curve_fit(multi_lorentz, xf, yf, p0=initial_guess)

# ✅ Plot
plt.figure(figsize=(14,5))
plt.plot(xf, yf, label="FFT-Spektrum", alpha=0.6)
plt.plot(xf, multi_lorentz(xf, *popt), 'r--', label="Lorentz-Fit")
plt.xlabel("Frequenz (ω)")
plt.ylabel("Amplitude")
plt.title("Lorentz-Fit auf FFT-Spektrum")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ✅ Ergebnis: Parameterausgabe
n_peaks = len(popt)//3
print("✅ Rekonstruierte Lorentz-Peaks:")
for i in range(n_peaks):
    A, w0, gamma = popt[3*i:3*i+3]
    print(f"Peak {i+1}: A = {A:.2f}, ω₀ = {w0:.5f}, γ = {gamma:.5f}")
```

```python
# Installation (nur für Google Colab)
# !pip install mpmath numba

import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero, diff, zeta, mp
from numba import njit, prange
import time

mp.dps = 40  # hohe Präzision für Zeta-Werte

# 1. Zeta-Nullstellen & Beta vorbereiten
num_zeros = 100  # anpassen je nach Leistung
gammas = np.array([float(zetazero(k).imag) for k in range(1, num_zeros + 1)])
rhos = 0.5 + 1j * gammas
beta = np.ones(num_zeros)
beta /= np.sum(beta)

# Optional: Näherung für zeta'(rho) vorberechnen
zeta_derivs = np.array([abs(diff(zeta, 0.5 + 1j * g)) for g in gammas])

# 2. Reale x-Werte
x_vals = np.linspace(10, 1000, 200)

# 3. GPU-fähige JIT-Rekonstruktion
@njit(parallel=True)
def reconstruct_L(x_vals, gammas, beta, zeta_derivs):
    Lx = np.zeros(x_vals.shape[0])
    for i in prange(x_vals.shape[0]):
        x = x_vals[i]
        sum_val = 0.0
        for k in range(len(gammas)):
            gamma = gammas[k]
            rho = 0.5 + 1j * gamma
            zprime = zeta_derivs[k]
            numerator = x**rho * beta[k]
            denominator = rho * zprime
            sum_val += (numerator / denominator).real
        Lx[i] = sum_val
    return Lx

# 4. Auswertung
start = time.time()
L_vals = reconstruct_L(x_vals, gammas, beta, zeta_derivs)
print(f"✅ Fertig in {time.time() - start:.2f} Sekunden")

# 5. Vergleich mit ψ(x)
from sympy import primerange
psi_x = [sum(np.log(p) for p in primerange(2, int(x)+1)) for x in x_vals]
errors = np.abs(L_vals - psi_x)

# 6. Plot
plt.figure(figsize=(10, 5))
plt.plot(x_vals, L_vals, label="L(x) GPU-optimiert", linewidth=2)
plt.plot(x_vals, psi_x, "--", label=r"$\psi(x)$")
plt.xlabel("x")
plt.ylabel("Wert")
plt.title("GPU-optimierte Rekonstruktion von ψ(x)")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(x_vals, errors, color='darkred')
plt.title("Fehler: |L(x) - ψ(x)|")
plt.xlabel("x")
plt.ylabel("Absoluter Fehler")
plt.grid()
plt.show()
```

```python
# OPTIONAL: Für Google Colab ausführen
# !pip install mpmath numba sympy

import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero, diff, zeta, mp
from numba import njit, prange
from sympy import primerange
import time

# 1. Genauigkeit & Setup
mp.dps = 40
num_zeros = 100  # Anzahl der Nullstellen
x_vals = np.linspace(10, 1000, 200)

# 2. Vorberechnung der Werte mit mpmath
print("Lade Zeta-Nullstellen & Ableitungen ...")
gammas = np.array([float(zetazero(k).imag) for k in range(1, num_zeros + 1)], dtype=np.float64)
zeta_derivs = np.array([abs(diff(zeta, 0.5 + 1j * g)) for g in gammas], dtype=np.float64)
beta = np.ones(num_zeros, dtype=np.float64)
beta /= np.sum(beta)

# 3. JIT-kompilierte, Numba-optimierte Rekonstruktion
@njit(parallel=True)
def reconstruct_L(x_vals, gammas, beta, zeta_derivs):
    Lx = np.zeros(x_vals.shape[0])
    for i in prange(x_vals.shape[0]):
        x = x_vals[i]
        sum_val = 0.0
        for k in range(len(gammas)):
            gamma = gammas[k]
            rho = complex(0.5, gamma)
            zprime = zeta_derivs[k]
            numerator = x**rho * beta[k]
            denominator = rho * zprime
            sum_val += (numerator / denominator).real
        Lx[i] = sum_val
    return Lx

# 4. Ausführung
start = time.time()
L_vals = reconstruct_L(x_vals, gammas, beta, zeta_derivs)
print(f"✅ Rekonstruktion abgeschlossen in {time.time() - start:.2f} Sekunden.")

# 5. Referenz: echte ψ(x) aus Primzahllogarithmen
psi_x = [sum(np.log(p) for p in primerange(2, int(x)+1)) for x in x_vals]
errors = np.abs(L_vals - psi_x)

# 6. Plot: L(x) vs. ψ(x)
plt.figure(figsize=(10, 5))
plt.plot(x_vals, L_vals, label="L(x) rekonstruiert", linewidth=2)
plt.plot(x_vals, psi_x, "--", label=r"$\psi(x)$ (Primzahlsumme)")
plt.xlabel("x")
plt.ylabel("Wert")
plt.title("Rekonstruktion von $\psi(x)$ aus Zeta-Nullstellen")
plt.legend()
plt.grid(True)
plt.show()

# 7. Plot: Fehlerverlauf
plt.figure(figsize=(10, 4))
plt.plot(x_vals, errors, color='crimson')
plt.title("Absoluter Fehler: $|L(x) - \psi(x)|$")
plt.xlabel("x")
plt.ylabel("Fehler")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero, diff, zeta, mp
from sympy.ntheory import mobius

# --- Einstellungen ---
mp.dps = 50                # Dezimalstellenpräzision
N = 1000                   # Länge der diskreten Skala
K = 50                     # Anzahl Zeta-Nullstellen (klein halten wegen Performance)
epsilon = 0.0              # RH-Verletzung (z. B. 0.01)
sigma_gauss = 0.5          # Breite der Gaußverteilung
lambda_mu = 0.05           # Stärke der Möbius-Rückkopplung

# --- Lade Nullstellen ---
gammas = [float(zetazero(k).imag) for k in range(1, K + 1)]
sigmas = [0.5 + epsilon for _ in range(K)]

# --- Berechne Gewichte: 1 / |zeta'(rho)| ---
weights = []
for sigma, gamma in zip(sigmas, gammas):
    rho = complex(sigma, gamma)
    zeta_prime = diff(zeta, rho)
    weights.append(1.0 / abs(zeta_prime))

# Optional: normiere Gewichte, um Überdominanz zu vermeiden
weight_sum = sum(weights)
weights = [w / weight_sum for w in weights]

# --- Testfunktionen ---
n_vals = np.arange(N)
f = np.sin(0.01 * np.pi * n_vals)
g = np.cos(0.01 * np.pi * n_vals)

# --- Operator Dµ(f) berechnen ---
Dmu_f = np.zeros_like(f)

# Spektralteil (Zeta-Nullstellen via Gauß-Delta)
for w, gamma in zip(weights, gammas):
    for n in range(N):
        delta = np.exp(-((n - gamma)**2) / (2 * sigma_gauss**2))
        Dmu_f[n] += w * delta * f[n]

# Möbius-Rückkopplungsteil (Spencer-Brown-Form)
for n in range(2, N):
    mu_sum = sum([mobius(d) * (n // d) for d in range(1, n)])
    Dmu_f[n] += lambda_mu * mu_sum * f[n]

# --- Adjungiertheit testen ---
inner1 = np.dot(Dmu_f, g)
inner2 = np.dot(f, Dmu_f)

print(f"⟨Dµ f, g⟩ = {inner1}")
print(f"⟨f, Dµ g⟩ = {inner2}")
print(f"Differenz: {abs(inner1 - inner2)}")

# --- Visualisierung ---
plt.figure(figsize=(10, 4))
plt.plot(n_vals, Dmu_f, label='Dµ(f)', color='darkred')
plt.plot(n_vals, f, '--', label='f(n)', color='steelblue')
plt.legend()
plt.title("Arithmetischer Dirac-Operator mit Möbius-Kopplung")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# --- Installation für lokale Umgebungen (nicht nötig in Colab) ---
# !pip install numpy matplotlib scipy

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# --- Parameter ---
file_path = '/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt'  # <- Pfad zu deiner Nullstellen-Datei
window_size = 5000  # Sliding Window Größe
step_size = 100     # Schrittweite
min_n = 10          # Untere Grenze für erste Fits
max_n = 100_000     # Obere Grenze

# --- 1. Zeta-Nullstellen laden ---
try:
    with open(file_path, 'r') as f:
        gamma = np.array([float(line.strip()) for line in f if line.strip()])
    print(f'✅ Geladene Nullstellen: {len(gamma)}')
except Exception as e:
    print(f'Fehler beim Einlesen: {e}')
    gamma = np.linspace(1, max_n + 1, max_n + 1)  # Fallback

# --- 2. Kohärenzlänge berechnen ---
L = gamma[1:] - gamma[:-1]

# --- 3. Fit-Funktion ---
def power_law(n, alpha, beta):
    return alpha * n**beta

# --- 4. Sliding Window Fit von Beta ---
fit_ns = []
betas = []

for start in range(min_n, max_n - window_size, step_size):
    end = start + window_size
    x = np.arange(start, end)
    y = L[start:end]

    try:
        popt, _ = curve_fit(power_law, x, y, p0=[4.0, 0.28])
        fit_ns.append(start + window_size // 2)
        betas.append(popt[1])
    except:
        continue

# --- 5. Visualisierung ---
plt.figure(figsize=(10, 5))
plt.plot(fit_ns, betas, 'b-', lw=2)
plt.axhline(np.mean(betas), color='r', linestyle='--', label='Mittelwert β')
plt.xlabel('Index $n$ (Mitte des Sliding Windows)')
plt.ylabel('Exponent $\\beta$')
plt.title('Sliding Window Fit von $L(n) \\approx \\alpha n^\\beta$')
plt.grid(True)
plt.legend()
plt.show()

# --- 6. Beta-Stabilitätsindex berechnen ---
beta_range = max(betas) - min(betas)
print(f"✅ Beta-Stabilitätsindex Δβ = {beta_range:.5f}")
```

```python
# ✅ Installation (nur beim ersten Mal nötig)
!pip install mpmath

# ✅ Imports
import numpy as np
import torch
import matplotlib.pyplot as plt
from mpmath import zeta, diff
from sympy import li, re

# ✅ Gerät auswählen
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("✅ Verwende Gerät:", device)

# ✅ Parameter
MAX_N = 20
x_vals = np.arange(1, 101)

# ✅ Lade rekonstruktive β-Skala (nur erste Spalte, Kopfzeile überspringen)
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta_vals = np.genfromtxt(beta_path, delimiter=",", skip_header=1, usecols=0, max_rows=MAX_N)
beta = torch.tensor(beta_vals, dtype=torch.float32, device=device)

# ✅ Dominante Frequenzen & Amplituden
frequencies = torch.tensor([
    9.994743e-06, 9.495006e-06, 8.995268e-06, 8.495531e-06, 7.995794e-06,
    7.496057e-06, 6.996320e-06, 6.496583e-06, 5.996846e-06, 5.497109e-06,
    4.997371e-06, 4.497634e-06, 3.997897e-06, 3.498160e-06, 2.998423e-06,
    2.498686e-06, 1.998949e-06, 9.994743e-07, 1.499211e-06, 4.997371e-07
], device=device)

amplitudes = torch.tensor([
    1.82e8, 1.91e8, 2.00e8, 2.11e8, 2.23e8, 2.36e8, 2.51e8, 2.68e8, 2.87e8, 3.09e8,
    3.35e8, 3.66e8, 4.03e8, 4.47e8, 5.02e8, 5.70e8, 6.54e8, 7.44e8, 7.45e8, 1.30e9
], device=device)

# ✅ Nullstellen γ (aus Frequenzen)
gamma = 2 * np.pi * frequencies.cpu().numpy()

# ✅ Formel (76): Liouville-Korrektur
def L_beta_weighted(beta, gamma, x_vals):
    beta = beta.to(device)
    gamma = gamma[:len(beta)]
    result = []
    for x in x_vals:
        x_log = torch.tensor(np.log(x), device=device)
        rho = torch.complex(torch.tensor(0.5), torch.tensor(gamma)).to(device)
        x_rho = torch.exp(x_log * rho)
        zeta_2rho = torch.tensor([complex(zeta(2 * r.item())) for r in rho], device=device)
        dzeta = torch.tensor([complex(diff(zeta, r.item())) for r in rho], device=device)
        term = beta * x_rho * zeta_2rho / (rho * dzeta)
        result.append(1 + term.real.sum().item())
    return np.array(result)

# ✅ Werte berechnen
L_vals = L_beta_weighted(beta, gamma, x_vals)

# ✅ Korrigierte Psi(x)-Funktion
def psi(x):
    return sum([re(li(x**p)) for p in range(1, 21)])

psi_vals = np.array([psi(x) for x in x_vals])

# ✅ Plot
plt.figure(figsize=(14, 6))
plt.plot(x_vals, psi_vals, label=r"$\psi(x)$ (klassisch)", color="black")
plt.plot(x_vals, L_vals, label=r"$L(x)$ mit $\beta_{\text{rek}}$ (20 Terme)", linestyle="--", color="blue")
plt.xlabel("x")
plt.ylabel("Wert")
plt.title("Vergleich: $\psi(x)$ vs. rekonstruktives $L(x)$ aus Frequenzformel")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange
from scipy.signal import find_peaks
from scipy.fft import fft, fftfreq

# ✅ Parameter
x_max = 100  # Intervallgrenze für x
dx = 0.01
x_vals = np.arange(1, x_max, dx)

# ✅ Primzahlen vorbereiten (bis z. B. 10^4)
max_prime = 10_000
primes = list(primerange(2, max_prime))

# ✅ Euler-Wellenfunktion berechnen
psi_x = np.zeros_like(x_vals)
for p in primes:
    psi_x += np.sin(x_vals * np.log(p))

# ✅ Normieren
psi_x /= np.sqrt(len(primes))

# ✅ Plot: Euler-Wellenfunktion
plt.figure(figsize=(12, 4))
plt.plot(x_vals, psi_x, color="darkblue", label=r"$\psi(x) = \sum \sin(x \log p)$")
plt.title("Euler-Wellenfunktion für Primzahlen bis {}".format(max_prime))
plt.xlabel("x")
plt.ylabel("ψ(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ✅ Fourier-Spektrum (Analyse dominanter Frequenzen)
spectrum = np.abs(fft(psi_x))
frequencies = fftfreq(len(x_vals), dx)

# ✅ Nur positive Frequenzen
mask = frequencies > 0
frequencies = frequencies[mask]
spectrum = spectrum[mask]

# ✅ Dominante Peaks
peaks, _ = find_peaks(spectrum, height=np.max(spectrum) * 0.1)
top_frequencies = frequencies[peaks]
top_amplitudes = spectrum[peaks]

# ✅ Ausgabe der Top-Frequenzen
sorted_idx = np.argsort(top_amplitudes)[-5:][::-1]
print("✅ Dominante Frequenzen:")
for i in sorted_idx:
    print(f"Frequenz f = {top_frequencies[i]:.5f} | Amplitude A = {top_amplitudes[i]:.2f}")

# ✅ Plot: Spektrum
plt.figure(figsize=(12, 4))
plt.plot(frequencies, spectrum, label="Spektrum von ψ(x)", color="crimson")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Euler-Wellenfunktion")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# ✅ Installationen (falls nötig)
!pip install scipy numpy matplotlib

# ✅ Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, fftfreq

# ✅ Lorentz-Funktion definieren
def lorentzian(omega, A, omega_0, gamma):
    return A * gamma**2 / ((omega - omega_0)**2 + gamma**2)

# ✅ Mehrere Lorentzians summieren
def multi_lorentz(omega, *params):
    n_peaks = len(params) // 3
    y = np.zeros_like(omega)
    for i in range(n_peaks):
        A = params[3*i]
        omega_0 = params[3*i+1]
        gamma = params[3*i+2]
        y += lorentzian(omega, A, omega_0, gamma)
    return y

# ✅ Beispielsignal: Summe von drei Sinusschwingungen
N = 10000
T = 1.0
t = np.linspace(0, N*T, N)
signal = (
    3.0 * np.sin(2 * np.pi * 0.25615 * t) +
    2.0 * np.sin(2 * np.pi * 0.3097 * t) +
    1.5 * np.sin(2 * np.pi * 0.45092 * t)
)

# ✅ FFT berechnen
yf = np.abs(fft(signal))[:N//2]
xf = fftfreq(N, T)[0:N//2]

# ✅ Anfangsschätzung für drei Lorentz-Peaks
initial_guess = [
    max(yf), 0.25615, 0.01,   # Peak 1
    max(yf)/2, 0.3097, 0.01,  # Peak 2
    max(yf)/3, 0.4509, 0.01   # Peak 3
]

# ✅ Lorentz-Fit durchführen
popt, _ = curve_fit(multi_lorentz, xf, yf, p0=initial_guess)

# ✅ Plot
plt.figure(figsize=(14,5))
plt.plot(xf, yf, label="FFT-Spektrum", alpha=0.6)
plt.plot(xf, multi_lorentz(xf, *popt), 'r--', label="Lorentz-Fit")
plt.xlabel("Frequenz (ω)")
plt.ylabel("Amplitude")
plt.title("Lorentz-Fit auf FFT-Spektrum")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# ✅ Ergebnis: Parameterausgabe
n_peaks = len(popt)//3
print("✅ Rekonstruierte Lorentz-Peaks:")
for i in range(n_peaks):
    A, w0, gamma = popt[3*i:3*i+3]
    print(f"Peak {i+1}: A = {A:.2f}, ω₀ = {w0:.5f}, γ = {gamma:.5f}")
```

```python
# ✅ Installation falls nötig
!pip install scipy matplotlib numpy

# ✅ Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.optimize import curve_fit

# ✅ Euler-Wellenfunktion definieren (für Primzahlen p ≤ p_max)
from sympy import primerange
p_max = 100
x_vals = np.linspace(0, 100, 10000)
primes = list(primerange(2, p_max + 1))

def euler_wave(x):
    return np.sum([np.sin(x * np.log(p)) for p in primes], axis=0)

psi_x = euler_wave(x_vals)

# ✅ FFT berechnen
N = len(x_vals)
dx = x_vals[1] - x_vals[0]
frequencies = fftfreq(N, dx)
spectrum = np.abs(fft(psi_x))

# ✅ Nur positive Frequenzen
freqs_pos = frequencies[:N // 2]
spectrum_pos = spectrum[:N // 2]

# ✅ Lorentzfunktion definieren
def lorentz(f, A, f0, gamma):
    return A / ((f - f0) ** 2 + gamma ** 2)

# ✅ Dominante Peaks auswählen und fitten
peak_indices = np.argsort(spectrum_pos)[-5:][::-1]
dominant_freqs = freqs_pos[peak_indices]
dominant_amps = spectrum_pos[peak_indices]

print("✅ Rekonstruierte Lorentz-Peaks:")
for i, idx in enumerate(peak_indices):
    f0_guess = freqs_pos[idx]
    A_guess = spectrum_pos[idx]
    gamma_guess = 0.0001

    # Fit in kleinem Frequenzfenster
    mask = (freqs_pos > f0_guess - 0.01) & (freqs_pos < f0_guess + 0.01)
    try:
        popt, _ = curve_fit(lorentz, freqs_pos[mask], spectrum_pos[mask], p0=[A_guess, f0_guess, gamma_guess])
        A_fit, f0_fit, gamma_fit = popt
        print(f"Peak {i+1}: A = {A_fit:.2f}, ω₀ = {f0_fit:.5f}, γ = {gamma_fit:.5f}")
    except RuntimeError:
        print(f"Peak {i+1}: Kein Fit möglich.")

# ✅ Plot der FFT
plt.figure(figsize=(12, 5))
plt.plot(freqs_pos, spectrum_pos, label="FFT |ψ(x)|", color="navy")
plt.scatter(dominant_freqs, dominant_amps, color="red", label="Dominante Peaks")
plt.xlabel("Frequenz f")
plt.ylabel("Amplitude")
plt.title("Fourier-Spektrum der Euler-Wellenfunktion")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === Pfad zu Odlyzko-Nullstellen ===
zeta_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
zeta_vals = pd.read_csv(zeta_path, header=None).iloc[:, 0].to_numpy()
N_vals = np.arange(1, len(zeta_vals) + 1)

# === Modellfunktion: 6 Parameter
def FFF_full(N, alpha, beta, gamma, delta, omega, phi):
    base = alpha * N**beta + gamma / N
    theta = delta * np.cos(omega * np.log(N + 1e-10) + phi)
    return base + theta

# === Fit durchführen (auf Subsample für Performance)
N_sample = N_vals[:2000]
zeta_sample = zeta_vals[:2000]

# === Startwerte (nahe notarielle Werte + moderate Theta-Startwerte)
p0 = [1.88, 0.91, 2488, 10000, 0.002, 0.0]
params, _ = curve_fit(FFF_full, N_sample, zeta_sample, p0=p0, maxfev=20000)
alpha_f, beta_f, gamma_f, delta_f, omega_f, phi_f = params

# === Ergebnisse anzeigen
print("✅ Optimierte Parameter:")
print(f"α (Basis):     {alpha_f:.6f}")
print(f"β (Exponent):  {beta_f:.6f}")
print(f"γ (Komp.):     {gamma_f:.6f}")
print(f"δ (Amplitude): {delta_f:.6f}")
print(f"ω (Frequenz):  {omega_f:.6f}")
print(f"φ (Phase):     {phi_f:.6f}")

# === Modellwerte auf kompletter Folge berechnen
zeta_fitted = FFF_full(N_vals, *params)

# === Fehlermaße berechnen
mae = np.mean(np.abs(zeta_vals - zeta_fitted))
mse = np.mean((zeta_vals - zeta_fitted)**2)

print(f"\n📉 MAE (6-Param-Fit): {mae:.6f}")
print(f"📉 MSE (6-Param-Fit): {mse:.6f}")

# === Plot
plt.figure(figsize=(10, 5))
plt.plot(N_vals[:300], zeta_vals[:300], label="Zeta-Nullstellen", linewidth=2)
plt.plot(N_vals[:300], zeta_fitted[:300], '--', label="FFF + Theta (6-Param)", linewidth=2)
plt.title("Zeta vs. vollständiger 6-Parameter-Fit")
plt.xlabel("Index N")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Primzahlen einlesen ===
prime_data = pd.read_csv('/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv',
                         header=None, dtype={0: np.int64})[0]
primes = prime_data.to_numpy()

# === Tschebyschow-Funktion ψ(x) berechnen ===
def psi_chebyshev(x_vals, primes):
    psi_vals = []
    for x in x_vals:
        total = 0
        for p in primes:
            if p > x:
                break
            power = p
            while power <= x:
                total += np.log(p)
                power *= p
        psi_vals.append(total)
    return np.array(psi_vals)

# === Wertebereich für x definieren ===
x_vals = np.arange(1, 1001)
psi_vals = psi_chebyshev(x_vals, primes)

# === Plot ===
plt.figure(figsize=(10, 5))
plt.plot(x_vals, psi_vals, label=r'$\psi(x)$ (klassisch)', color='green')
plt.title("Tschebyschow-Funktion ψ(x)")
plt.xlabel("x")
plt.ylabel(r"$\psi(x)$")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# === Parameter aus der erweiterten FFF (aus Notiz)
A = 1.83282
B = 0.918
C = 2.488144455
D = 0.0055
delta = -8840.034295
omega = 0.001512
phi = 1.569544

# === Erweiterte FFF-Formel
def FFF_extended(N):
    N = np.array(N, dtype=np.float64)
    return (
        A * N**B +
        C * np.log(N + 1e-9) +
        D / (N**0.318) +
        delta * np.sin(omega * np.log(N + 1e-9) + phi)
    )

# === Berechnung
N_vals = np.arange(2, 1001)
L_vals = FFF_extended(N_vals)

# === Primzahlen einlesen
prime_data = pd.read_csv('/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.csv',
                         header=None, comment='P', dtype=str)[0]
primes = pd.to_numeric(prime_data, errors='coerce').dropna().astype(int).to_numpy()

# === π(N): Anzahl Primzahlen bis N
pi_vals = np.array([np.sum(primes <= n) for n in N_vals])

# === Fehler berechnen
abs_diff = np.abs(L_vals - pi_vals)

# === Plots
plt.figure(figsize=(12, 4))

# --- Plot: Funktion vs. Primzahlen
plt.subplot(1, 2, 1)
plt.plot(N_vals, L_vals, label='Erweiterte FFF', color='blue')
plt.plot(N_vals, pi_vals, label='π(N): Primzahlanzahl', color='green', linestyle='--')
plt.xlabel('N')
plt.ylabel('Wert')
plt.title('Vergleich: Erweiterte FFF vs. Primzahlanzahl')
plt.legend()

# --- Plot: Fehlerverlauf
plt.subplot(1, 2, 2)
plt.plot(N_vals, abs_diff, label='Abweichung |FFF - π(N)|', color='red')
plt.xlabel('N')
plt.ylabel('Abweichung')
plt.title('Fehlerverlauf der erweiterten FFF')
plt.legend()

plt.tight_layout()
plt.show()
```
