```python
# ğŸ“Œ Bibliotheken importieren
!pip install numpy pandas matplotlib seaborn scipy tqdm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from scipy.optimize import minimize
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# ğŸ“Œ Zeta-Nullstellen laden & normalisieren
zeta_file = "/content/drive/MyDrive/zeros6.txt"
zeta_zeros = np.loadtxt(zeta_file)[:100000]  # Begrenzung auf 100k Nullstellen
scaler = StandardScaler()
zeta_zeros_norm = scaler.fit_transform(zeta_zeros.reshape(-1, 1)).flatten()

# ğŸ“Œ Funktion zur Bewertung eines Beta-Werts
def evaluate_beta(beta):
    predicted_values = beta * np.ones_like(zeta_zeros_norm)  # Konstante Werte fÃ¼r Beta
    r2 = r2_score(zeta_zeros_norm, predicted_values)  # RÂ² berechnen
    chi2, p = chi2_contingency(np.array([zeta_zeros_norm, predicted_values]))[:2]  # ChiÂ²-Test
    return r2, chi2, p

# ğŸ“Œ Beta-Werte von 1e-6 bis 1e-1 testen
beta_test_values = np.logspace(-6, -1, 1000)
best_beta, best_chi2 = None, np.inf
results = []

print("\nğŸš€ Suche nach dem besten Beta-Wert...")
for beta in tqdm(beta_test_values):
    r2, chi2, p = evaluate_beta(beta)
    results.append((beta, r2, chi2, p))
    if chi2 < best_chi2:
        best_beta, best_chi2 = beta, chi2

# ğŸ“Œ DataFrame fÃ¼r Ergebnisse
df_results = pd.DataFrame(results, columns=["Beta-Wert", "RÂ²", "ChiÂ²", "p-Wert"])

# ğŸ“Œ Bestes Beta ausgeben
print(f"\nâœ… Bester gefundener Beta-Wert: {best_beta:.10f} mit ChiÂ² = {best_chi2:.4f}")

# ğŸ“Œ Visualisierung der ChiÂ²-Minimierung
plt.figure(figsize=(8, 5))
plt.plot(beta_test_values, df_results["ChiÂ²"], label="ChiÂ²-Wert")
plt.axvline(x=best_beta, color="red", linestyle="dashed", label=f"Bestes Beta = {best_beta:.8f}")
plt.xscale("log")
plt.xlabel("Beta-Wert (log-Skala)")
plt.ylabel("ChiÂ²-Fehler")
plt.title("Optimierung des besten Beta-Werts")
plt.legend()
plt.grid()
plt.show()
```
