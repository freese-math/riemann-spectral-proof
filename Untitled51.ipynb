```python
# 📌 Bibliotheken importieren
!pip install numpy pandas matplotlib seaborn scipy tqdm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
from scipy.optimize import minimize
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

# 📌 Zeta-Nullstellen laden & normalisieren
zeta_file = "/content/drive/MyDrive/zeros6.txt"
zeta_zeros = np.loadtxt(zeta_file)[:100000]  # Begrenzung auf 100k Nullstellen
scaler = StandardScaler()
zeta_zeros_norm = scaler.fit_transform(zeta_zeros.reshape(-1, 1)).flatten()

# 📌 Funktion zur Bewertung eines Beta-Werts
def evaluate_beta(beta):
    predicted_values = beta * np.ones_like(zeta_zeros_norm)  # Konstante Werte für Beta
    r2 = r2_score(zeta_zeros_norm, predicted_values)  # R² berechnen
    chi2, p = chi2_contingency(np.array([zeta_zeros_norm, predicted_values]))[:2]  # Chi²-Test
    return r2, chi2, p

# 📌 Beta-Werte von 1e-6 bis 1e-1 testen
beta_test_values = np.logspace(-6, -1, 1000)
best_beta, best_chi2 = None, np.inf
results = []

print("\n🚀 Suche nach dem besten Beta-Wert...")
for beta in tqdm(beta_test_values):
    r2, chi2, p = evaluate_beta(beta)
    results.append((beta, r2, chi2, p))
    if chi2 < best_chi2:
        best_beta, best_chi2 = beta, chi2

# 📌 DataFrame für Ergebnisse
df_results = pd.DataFrame(results, columns=["Beta-Wert", "R²", "Chi²", "p-Wert"])

# 📌 Bestes Beta ausgeben
print(f"\n✅ Bester gefundener Beta-Wert: {best_beta:.10f} mit Chi² = {best_chi2:.4f}")

# 📌 Visualisierung der Chi²-Minimierung
plt.figure(figsize=(8, 5))
plt.plot(beta_test_values, df_results["Chi²"], label="Chi²-Wert")
plt.axvline(x=best_beta, color="red", linestyle="dashed", label=f"Bestes Beta = {best_beta:.8f}")
plt.xscale("log")
plt.xlabel("Beta-Wert (log-Skala)")
plt.ylabel("Chi²-Fehler")
plt.title("Optimierung des besten Beta-Werts")
plt.legend()
plt.grid()
plt.show()
```

```python
# 📌 Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from scipy.fftpack import fft

# 📌 Dateipfad (Zeta-Nullstellen)
zeta_file = "/content/drive/MyDrive/zeros6.txt"

# 📌 Nullstellen laden
print("🔄 Lade 2 Mio Zeta-Nullstellen...")
zeta_zeros = np.loadtxt(zeta_file)
zeta_zeros = zeta_zeros[:2000000]  # Falls Datei größer ist

# 📌 Fourier-Analyse der Zeta-Nullstellen
print("🔍 Fourier-Transformation...")
fourier_spectrum = np.abs(fft(zeta_zeros))[:len(zeta_zeros)//2]

# 📌 Logarithmische Modifikation der Beta-Funktion
def beta_mod(N, beta_0, c):
    return beta_0 + c / np.log(N)

# 📌 Fehlerfunktion zur Optimierung
def error_function(c, N_values, beta_0):
    beta_values = beta_mod(N_values, beta_0, c)
    error = np.abs(np.exp(1j * beta_values * np.pi) + 1)  # Euler-Freese-Abweichung
    return np.mean(error**2)  # MSE minimieren

# 📌 Werte für N definieren
N_values = np.linspace(10, len(zeta_zeros), 500)  # 500 Werte für N

# 📌 Startwert von Beta
beta_0 = 1/137  # Referenzwert

# 📌 Optimierung von c
print("🚀 Optimiere Beta-Wert...")
result = minimize(error_function, x0=0.1, args=(N_values, beta_0), method='Nelder-Mead')
best_c = result.x[0]

# 📌 Berechnung der optimierten Beta-Werte
beta_values_optimized = beta_mod(N_values, beta_0, best_c)

# 📌 Visualisierung der optimierten Beta-Werte
plt.figure(figsize=(10, 5))
plt.plot(N_values, beta_values_optimized, label=r'Optimierte $\beta(N)$', color='blue', linewidth=2)
plt.axhline(1/137, linestyle="--", color="red", label=r"$\beta = \frac{1}{137}$")
plt.axhline(7/33300, linestyle="--", color="green", label=r"$\beta = \frac{7}{33300}$")
plt.xscale("log")
plt.xlabel("N")
plt.ylabel("Beta-Wert")
plt.title("Optimierung von Beta mit Zeta-Nullstellen")
plt.legend()
plt.show()

# 📌 Ergebnisse ausgeben
print(f"✅ Optimierter Wert von c: {best_c:.10f}")
```

```python
# 📌 Google Drive mounten
from google.colab import drive
drive.mount('/content/drive')

# 📌 Wichtige Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import scipy.fftpack
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from scipy.stats import chi2_contingency

# 📌 Nullstellen aus Datei laden
file_path = "/content/drive/MyDrive/zeros6.txt"
zeta_zeros = np.loadtxt(file_path)[:2000000]  # Nutze 2 Mio Nullstellen

# 📌 Fourier-Analyse der Nullstellen
fourier_spectrum = np.abs(scipy.fftpack.fft(zeta_zeros))[:len(zeta_zeros)//2]

# 📌 Modulo-Features erstellen
modulo_features = np.mod(zeta_zeros, np.e)

# 📌 Feature-Matrix & Zielvariable
X = np.vstack([zeta_zeros, fourier_spectrum[:len(zeta_zeros)], modulo_features]).T
y = np.full(len(X), 1/137)  # Startwert für Beta

# 📌 Daten normalisieren & splitten
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 📌 Neuronales Netz für Beta-Vorhersage
class BetaNet(nn.Module):
    def __init__(self):
        super(BetaNet, self).__init__()
        self.fc1 = nn.Linear(X.shape[1], 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.fc3(x)

# 📌 Modell, Loss & Optimizer
model = BetaNet()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 📌 Training auf CPU
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.reshape(-1, 1), dtype=torch.float32)

# 📌 Training starten
epochs = 50
losses = []
for epoch in tqdm(range(epochs), desc="🚀 ML-Training läuft"):
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()
    losses.append(loss.item())

    if epoch % 10 == 0:
        print(f"📊 Epoch {epoch}/{epochs} - Loss: {loss.item():.8f}")

# 📌 Vorhersage & Bewertung
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_pred_tensor = model(X_test_tensor).detach().numpy()

# 📌 Statistische Bewertung
r2 = np.corrcoef(y_test, y_pred_tensor.flatten())[0, 1] ** 2
chi2, p_value = chi2_contingency(np.histogram2d(y_test, y_pred_tensor.flatten(), bins=10)[0])[:2]

# 📌 Euler-Freese-Test
def euler_freese_test(beta):
    return np.abs(2 - (beta * np.pi)**2)

best_beta = np.mean(y_pred_tensor)
euler_error = euler_freese_test(best_beta)

# 📌 Ergebnisplot
plt.figure(figsize=(8, 5))
plt.bar(["1/137", "1/129.4", "7/33300", "ML-Predicted"],
        [euler_freese_test(1/137), euler_freese_test(1/129.4), euler_freese_test(7/33300), euler_error],
        color=["gray", "gray", "gray", "yellow"])
plt.ylabel("Euler-Freese-Abweichung")
plt.title("Highscore-Optimierung für Beta")
plt.legend(["Beta-Werte"])
plt.show()

# 📌 Endgültiges Ergebnis ausgeben
print("\n🎯 **Ergebnisse:**")
print(f"✅ ML abgeschlossen! R² = {r2:.6f}, Chi² = {chi2:.6f}, p-Wert = {p_value:.6f}")
print(f"🏆 **Bester vorhergesagter Beta-Wert:** {best_beta:.10f}")
print(f"✅ Euler-Freese Abweichung (ML-Beta) = {euler_error:.10f}")
```
