```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.functions.combinatorial.numbers import mobius
from sympy import totient
from mpmath import zetazero, zeta, diff, mp

mp.dps = 30  # Przision fr '()

#  Parameter
N = 100      # Anzahl der Punkte (n = 1 bis N)
K = 20       # Anzahl der Zeta-Nullstellen fr die Spektralkomponenten
sigma = 2.0
lambda_mu = 0.1
n_vals = np.arange(1, N+1)

#  Zeta-Nullstellen (Imaginrteile) & Normierung der Gewichte
gammas = [float(zetazero(k).imag) for k in range(1, K + 1)]
weights = [1.0 / abs(diff(zeta, 0.5 + 1j * g)) for g in gammas]
weights = [w / sum(weights) for w in weights]

#  Gau-Delta-Funktion
def delta_sigma(n, gamma_k, sigma):
    return np.exp(-((n - gamma_k) ** 2) / (2 * sigma ** 2))

#  Operator D als Funktion, die auf f(n) wirkt
def D_mu_apply(f_vals):
    result = np.zeros_like(f_vals)
    for n in range(1, N+1):
        spectral = sum(weights[k] * delta_sigma(n, gammas[k], sigma) for k in range(K))
        arithmetic = lambda_mu * sum(mobius(d) * (n // d) for d in range(1, n))
        result[n-1] = (spectral + arithmetic) * f_vals[n-1]
    return result

#  Komponenten des Vierervektors P^(n)
# Komponente 0: Spektrale Energie (Platzhalter, z.B. log(n)/n)
beta = np.log(n_vals) / n_vals

# Komponente 1: Arithmetischer Impuls 1  Mbius / n
pi1 = np.array([mobius(n) / n for n in n_vals])

# Komponente 2: Arithmetischer Impuls 2  1/(n)
pi2 = np.array([1 / totient(n) if totient(n) != 0 else 0 for n in n_vals])

# Komponente 3: Gaudelta gegen Zeta-Imaginrteile
# Fr n <= K: benutze gamma[n-1], ansonsten den letzten Wert gammas[-1]
pi3 = np.array([np.exp(-((n - (gammas[n-1] if n <= len(gammas) else gammas[-1]))**2) / (2 * sigma**2)) for n in n_vals])

#  Gesamt-Vierervektor P^(n) (als 4 x N Matrix)
P_mu = np.vstack((beta, pi1, pi2, pi3))

#  Wirkung des Operators D auf jede Komponente
Dbeta = D_mu_apply(beta)
Dpi1  = D_mu_apply(pi1)
Dpi2  = D_mu_apply(pi2)
Dpi3  = D_mu_apply(pi3)

#  Ergebnis-Vierervektor nach Wirkung: D(P^(n))
DP_mu = np.vstack((Dbeta, Dpi1, Dpi2, Dpi3))

#  Plot: Vergleich der Originalkomponenten und der transformierten Komponenten
fig, axs = plt.subplots(4, 1, figsize=(10, 12), sharex=True)

axs[0].plot(n_vals, beta, 'b-', label=r"$\beta(n)$ (Original)")
axs[0].plot(n_vals, Dbeta, 'r--', label=r"$\mathcal{D}_\mu(\beta(n))$")
axs[0].legend()
axs[0].set_ylabel("Energie")

axs[1].plot(n_vals, pi1, 'b-', label=r"$\mu(n)/n$ (Original)")
axs[1].plot(n_vals, Dpi1, 'r--', label=r"$\mathcal{D}_\mu(\pi_1(n))$")
axs[1].legend()
axs[1].set_ylabel("Impuls 1")

axs[2].plot(n_vals, pi2, 'b-', label=r"$1/\varphi(n)$ (Original)")
axs[2].plot(n_vals, Dpi2, 'r--', label=r"$\mathcal{D}_\mu(\pi_2(n))$")
axs[2].legend()
axs[2].set_ylabel("Impuls 2")

axs[3].plot(n_vals, pi3, 'b-', label=r"$\delta_\sigma(n - \gamma(n))$ (Original)")
axs[3].plot(n_vals, Dpi3, 'r--', label=r"$\mathcal{D}_\mu(\pi_3(n))$")
axs[3].legend()
axs[3].set_ylabel("Impuls 3")
axs[3].set_xlabel("n")

plt.suptitle(r"Wirkung von $\mathcal{D}_\mu$ auf den arithmetisch-relativistischen Vierervektor $\mathcal{P}^\mu(n)$")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
from scipy.fft import fft, fftfreq

#  Zeta-Nullstellen laden
with open("/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt", "r") as f:
    gamma_vals = np.array([float(line.strip()) for line in f if line.strip()])

#  Beta-Werte laden
beta_df = pd.read_csv("/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv")
beta_vals = beta_df.iloc[:, 0].values

#  Begrenze auf gemeinsamen Bereich
N = min(len(beta_vals), len(gamma_vals))
beta_n = beta_vals[:N]
gamma_n = gamma_vals[:N]

#  Lineare Regression:   a + b
slope, intercept, r_value, _, _ = linregress(beta_n, gamma_n)
gamma_fit = slope * beta_n + intercept
error = gamma_n - gamma_fit

#  Plot: Fehlerstruktur (n)
plt.figure(figsize=(8, 4))
plt.plot(range(1, N+1), error, "g-", linewidth=1)
plt.title("Fehlerstruktur: $\\epsilon_n = \\gamma_n - (a \\cdot \\beta(n) + b)$")
plt.xlabel("n")
plt.ylabel("Abweichung (n)")
plt.grid(True)
plt.tight_layout()
plt.show()

#  FFT der Fehlerstruktur
fft_error = fft(error)
freqs = fftfreq(N)

plt.figure(figsize=(8, 4))
plt.plot(freqs[:N//2], np.abs(fft_error[:N//2]), "r-")
plt.title("Fourier-Spektrum der Fehlerstruktur")
plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.grid(True)
plt.tight_layout()
plt.show()

#  Inversion: _approx() = ( - b) / a
beta_inv = (gamma_n - intercept) / slope

#  Plot: Original vs. Inverse (n)
plt.figure(figsize=(8, 5))
plt.plot(range(1, N+1), beta_n, "b-", label=r"$\beta(n)$ original")
plt.plot(range(1, N+1), beta_inv, "r--", label=r"$\beta_{\mathrm{inv}}(\gamma_n)$")
plt.title("Vergleich: Originale und aus $\gamma_n$ rekonstruierte $\beta(n)$")
plt.xlabel("n")
plt.ylabel("Wert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  Korrelation der Inversion
corr_inv = np.corrcoef(beta_n, beta_inv)[0, 1]
print(f" Inverse Rekonstruktion erfolgreich!")
print(f" Pearson-Korrelation  vs. (): r = {corr_inv:.6f}")
```

```python
#  Google Drive mounten
from google.colab import drive
drive.mount('/content/drive')

#  Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange, log as symlog

#  Nullstellen laden
with open('/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt', 'r') as f:
    gammas = np.array([float(line.strip()) for line in f if line.strip()])

#  Primzahlen laden
with open('/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt', 'r') as f:
    primes = np.array([int(p) for p in f.read().split(',') if p.strip().isdigit()])

#  Werte fr x
x_vals = np.logspace(2.5, 5.2, 40, dtype=int)

#  Speicher fr E(x)
errors = []

for x in x_vals:
    # (x): Chebyshev-Funktion (mit allen Potenzen p^k <= x)
    psi_x = 0
    for p in primes:
        if p > x:
            break
        k = 1
        while (p**k <= x):
            psi_x += symlog(p)
            k += 1

    # R(x): Korrekturterm ber Nullstellen
    R_x = np.sum(np.exp(1j * gammas * np.log(x)) / (1j * gammas))
    Re_R_x = R_x.real

    # Fehlerfunktion
    E_x = psi_x - x + Re_R_x
    errors.append(E_x)

#  Plot
plt.figure(figsize=(10, 5))
plt.plot(x_vals, errors, 'o-', label=r"$\psi(x) - x + R(x)$")
plt.axhline(0, color='gray', linestyle='--')
plt.xscale('log')
plt.xlabel("x (log)")
plt.ylabel("Fehlerterm $E(x)$")
plt.title("RH-Test: Primzahldichte vs. Nullstellenstruktur")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  Nullstellen laden (nur Imaginrteile)
gammas = np.loadtxt("/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt")

#  Primzahlen laden und (x) = Sum_{p^k  x} log p berechnen
with open("/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt") as f:
    primes = list(map(int, f.read().split()))

def psi(x):
    return sum(np.log(p) for p in primes if p <= x)

#  RH-Korrekturterm: R(x) = sum(x^ / ) + konjugiert
def R(x, gammas, N=1000):
    total = 0.0
    for g in gammas[:N]:
        rho = 0.5 + 1j * g
        term = x**rho / rho + x**(1 - rho) / (1 - rho)
        total += term.real
    return total

#  Testpunkte (logarithmisch verteilt)
x_vals = np.logspace(2, 5, 40)
errors = []

#  Fehlerterm berechnen: E(x) = (x) - x + R(x)
for x in x_vals:
    px = psi(x)
    rx = R(x, gammas)
    ex = px - x + rx
    errors.append(ex)

#  Plot
plt.figure(figsize=(12, 5))
plt.plot(x_vals, errors, 'o-', label=r"$\psi(x) - x + R(x)$")
plt.axhline(0, color='gray', linestyle='--')
plt.xscale('log')
plt.xlabel("x (log)")
plt.ylabel("Fehlerterm $E(x)$")
plt.title("RH-Test: Primzahldichte vs. Nullstellenstruktur")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#  Notwendige Bibliotheken installieren (nur beim ersten Start)
# !pip install numpy pandas matplotlib

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  Dateien laden  bitte vorher in Colab hochladen oder mit Google Drive verbinden
zeros_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6.txt"
beta_path = "/content/drive/MyDrive/beta_skala_rekonstruiert_2M.csv"
prime_path = "/content/drive/MyDrive/Aus Chrome gespeichert/primzahlen2mio.txt"

#  Laden der echten Zeta-Nullstellen (nur Imaginrteile )
with open(zeros_path, "r") as f:
    gamma = np.array([float(line.strip()) for line in f.readlines() if line.strip()])

#  Laden der rekonstruierten Beta-Werte
beta = pd.read_csv(beta_path, header=None).iloc[:, 0].to_numpy()

#  Krzen auf gemeinsame Lnge
N = min(len(gamma), len(beta), 5000)
gamma = gamma[:N]
beta = beta[:N]

#  Rekonstruktive Spektralformel L(x)
def L_reconstruct(x, gamma, beta):
    rho = 0.5 + 1j * gamma
    numerator = beta * x**rho
    denominator = rho * (np.log(x + 1e-10) + 1)  # Nherung fr '()
    return np.real(np.sum(numerator / denominator))

#  x-Werte vorbereiten
x_vals = np.linspace(10, 1000, 500)
L_vals = np.array([L_reconstruct(x, gamma, beta) for x in x_vals])

#  Vergleich: Echte (x) aus Primzahlen aufbauen
def psi_true(x, primes):
    return np.sum([np.log(p) for p in primes if p <= x])

with open(prime_path, "r") as f:
    primes = [int(line.strip()) for line in f if line.strip()]
    primes = [p for p in primes if p <= 1000]

psi_vals = np.array([psi_true(x, primes) for x in x_vals])

#  Plot: Vergleich L(x) vs. (x)
plt.figure(figsize=(10, 5))
plt.plot(x_vals, psi_vals, label="(x) (echt)", color="green")
plt.plot(x_vals, L_vals, label="L(x) (rekonstruiert)", color="blue", linestyle="--")
plt.title("Vergleich: Rekonstruktive Spektralformel vs. echte Tschebyschow-Funktion")
plt.xlabel("x")
plt.ylabel("Funktionswert")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  Korrelation ausgeben
correlation = np.corrcoef(L_vals, psi_vals)[0, 1]
print(f" Pearson-Korrelation: r = {correlation:.6f}")
```
