```python
# ✅ Installation (nur einmal nötig)
!pip install sympy

# ✅ Imports
import numpy as np
import torch
import matplotlib.pyplot as plt
from sympy import mobius, divisor_sigma, primerange

# ✅ Gerät
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("✅ Verwende Gerät:", device)

# ✅ Parameter
MAX_N = 100
n_vals = torch.arange(1, MAX_N + 1, device=device)

# ✅ Rekonstruktion der Primzahlstruktur über zwei klassische arithmetische Funktionen:

# λ(n) = Liouville-Funktion (mit Vorzeichen der Primfaktoranzahl)
def liouville(n):
    def omega(n):
        p = list(primerange(1, n + 1))
        count = 0
        for q in p:
            while n % q == 0:
                count += 1
                n = n // q
        return count
    return (-1)**omega(n)

# η(n) = Dirichlet-Eta-Verschränkung via Möbius + Sigma (vereinfachtes Beispiel)
eta_vals = torch.tensor([mobius(int(n.item())) * divisor_sigma(1, int(n.item())) for n in n_vals], dtype=torch.float32, device=device)
lambda_vals = torch.tensor([liouville(int(n.item())) for n in n_vals], dtype=torch.float32, device=device)

# ✅ Plot
plt.figure(figsize=(10, 5))
plt.plot(n_vals.cpu(), eta_vals.cpu(), label=r"$\eta(n)$", linewidth=1.5)
plt.plot(n_vals.cpu(), lambda_vals.cpu(), label=r"$\lambda(n)$", linestyle="--", linewidth=1.5)
plt.xlabel("n")
plt.ylabel("Wert")
plt.title(r"Rekonstruierte Primzahlverteilung über $\eta(n)$ und $\lambda(n)$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ✅ Imports & Setup
import torch
import numpy as np
import matplotlib.pyplot as plt

# ✅ Gerät wählen
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("✅ Verwende Gerät:", device)

# ✅ Parameter
N = 10000
x = torch.arange(1, N + 1, device=device)

# ✅ Rekonstruktive Struktur (Drift + Frequenzen)
frequencies = torch.tensor([
    9.994743e-06, 9.495006e-06, 8.995268e-06, 8.495531e-06, 7.995794e-06,
    7.496057e-06, 6.996320e-06, 6.496583e-06, 5.996846e-06, 5.497109e-06,
    4.997371e-06, 4.497634e-06, 3.997897e-06, 3.498160e-06, 2.998423e-06,
    2.498686e-06, 1.998949e-06, 9.994743e-07, 1.499211e-06, 4.997371e-07
], device=device)

amplitudes = torch.tensor([
    1.82e8, 1.91e8, 2.00e8, 2.11e8, 2.23e8, 2.36e8, 2.51e8, 2.68e8, 2.87e8, 3.09e8,
    3.35e8, 3.66e8, 4.03e8, 4.47e8, 5.02e8, 5.70e8, 6.54e8, 7.44e8, 7.45e8, 1.30e9
], device=device)

# ✅ Drift
drift = 1.78e-8 * x**2 - 0.375 * x + 5.82e5

# ✅ ε(n) aus Frequenzüberlagerung
epsilon = torch.zeros_like(x, dtype=torch.float64)
for f, A in zip(frequencies, amplitudes):
    epsilon += A * torch.cos(2 * np.pi * f * x)

# ✅ Beta-Rekonstruktion
beta = epsilon - drift

# ✅ Signalverstärkung & Normierung
signal = torch.abs(beta)
signal = (signal - signal.min()) / (signal.max() - signal.min())

# ✅ Schwellenwert wählen & Primzahlpositionen vorhersagen
threshold = 0.99
predicted_primes = x[signal > threshold].cpu().numpy()

# ✅ Plot
plt.figure(figsize=(14, 5))
plt.plot(x.cpu(), signal.cpu(), label="Spektrales β-Signal", color="blue")
plt.scatter(predicted_primes, [1]*len(predicted_primes), color="red", s=10, label="Vorhergesagte Primzahlen")
plt.axhline(threshold, color="gray", linestyle="--", label="Schwelle")
plt.xlabel("n")
plt.ylabel("Signal")
plt.title("Spektrale Vorhersage von Primstellen mit rekonstruktivem β(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ✅ Installation (falls nötig)
!pip install sympy

# ✅ Imports
import numpy as np
import matplotlib.pyplot as plt
from sympy import isprime

# ✅ Parameter
N = 10000
n = np.arange(1, N + 1)

# ✅ Beispielhaftes rekonstruktives β(n)-Signal
beta_signal = 1 - (n / N) ** 2  # Platzhalter zur Veranschaulichung

# ✅ Schwellenwert
threshold = 0.98
predicted_primes = np.where(beta_signal > threshold)[0] + 1  # +1 wegen Indexstart

# ✅ Echte Primzahlen
true_primes_set = set(i for i in range(1, N + 1) if isprime(i))

# ✅ Genauigkeitsprüfung für verschiedene Bereiche
def prime_prediction_stats(predicted, true_set, limit):
    preds = set(predicted[predicted <= limit])
    true = set(filter(lambda x: x <= limit, true_set))
    hits = preds & true
    precision = len(hits) / len(preds) if preds else 0
    recall = len(hits) / len(true) if true else 0
    return len(preds), len(true), len(hits), precision, recall

limits = [100, 1000, 10000]
print("| Bereich | Vorhergesagte | Echte | Treffer | Precision | Recall |")
print("|---------|----------------|-------|---------|-----------|--------|")

for limit in limits:
    vp, ep, hit, p, r = prime_prediction_stats(predicted_primes, true_primes_set, limit)
    print(f"| ≤{limit:<5} | {vp:<14} | {ep:<5} | {hit:<7} | {p:<9.2f} | {r:<6.2f} |")

# ✅ Visualisierung
plt.figure(figsize=(14, 5))
plt.plot(n, beta_signal, label="Spektrales β-Signal", color="blue")
plt.axhline(threshold, linestyle="--", color="gray", label="Schwelle")
plt.scatter(predicted_primes, [1]*len(predicted_primes), color="red", label="Vorhergesagte Primzahlen", s=10)
plt.xlabel("n")
plt.ylabel("Signal")
plt.title("Spektrale Vorhersage von Primstellen mit rekonstruktivem β(n)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import isprime

# ✅ Beta-Signal vorbereiten (z. B. wie im letzten Plot)
N = 1000
n_vals = np.arange(N)
# Beispiel: P_beta[n] = normiertes spektrales Signal aus vorheriger Rekonstruktion
P_beta = np.load("/content/drive/MyDrive/beta_spektralrekonstruktion.npy")[:N]

# ✅ Ground truth: echte Primzahlen
P_true = np.array([1 if isprime(n) else 0 for n in n_vals])

# ✅ Schwellenwerte durchtesten
thresholds = np.linspace(0.1, 1.0, 100)
precision = []
recall = []

for T in thresholds:
    prediction = (P_beta > T).astype(int)
    TP = np.sum((prediction == 1) & (P_true == 1))
    FP = np.sum((prediction == 1) & (P_true == 0))
    FN = np.sum((prediction == 0) & (P_true == 1))
    prec = TP / (TP + FP + 1e-9)
    rec = TP / (TP + FN + 1e-9)
    precision.append(prec)
    recall.append(rec)

# ✅ Precision-Recall-Plot
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, marker="o")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Schwellenoptimierung: Precision vs. Recall")
plt.grid(True)
plt.tight_layout()
plt.show()

# ✅ Beste Schwelle (maximales F1)
f1_scores = 2 * np.array(precision) * np.array(recall) / (np.array(precision) + np.array(recall) + 1e-9)
best_idx = np.argmax(f1_scores)
best_T = thresholds[best_idx]
print(f"Beste Schwelle: T = {best_T:.3f} mit F1 = {f1_scores[best_idx]:.3f}")
```
