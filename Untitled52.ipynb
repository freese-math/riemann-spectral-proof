```python
# 📌 Notwendige Bibliotheken importieren
import numpy as np
import mpmath as mp
import scipy.optimize as opt
import matplotlib.pyplot as plt
from tqdm import tqdm

# 📌 Dezimalgenauigkeit erhöhen
mp.dps = 50

# 📌 Euler-Freese-Funktion
def euler_freese(nullstellen, beta):
    beta = mp.mpf(beta)
    return abs(mp.fsum([1 / mp.zeta(n + beta) for n in nullstellen]))

# 📌 Nullstellen aus Datei einlesen (mit Fehlerprüfung)
def load_zetas(file_path):
    valid_zeros = []
    try:
        with open(file_path, "r") as f:
            for line in f:
                try:
                    val = mp.mpf(line.strip())  # Hochpräzise Umwandlung
                    valid_zeros.append(val)
                except ValueError:
                    print(f"⚠️ Ungültiger Wert übersprungen: {line.strip()}")
    except Exception as e:
        print(f"❌ Fehler beim Laden der Datei: {e}")
        return None  # Falls Datei nicht geladen werden kann

    if not valid_zeros:
        print("❌ Keine gültigen Nullstellen gefunden! Fallback-Daten werden verwendet.")
        return np.linspace(14, 100, 1000)

    return np.array(valid_zeros, dtype=object)  # `dtype=object` für `mp.mpf`

# 📌 Hochpräzise Optimierung für β mit Fehlerbehandlung
def optimize_beta(nullstellen):
    try:
        result = opt.minimize_scalar(
            lambda beta: abs(euler_freese(nullstellen, beta) - 1000),
            bounds=(0.45, 0.50),
            method="bounded"
        )
        return result.x, result.fun
    except Exception as e:
        print(f"❌ Fehler in der Optimierung: {e}")
        return None, None

# 📌 Nullstellen laden
file_path = "/content/drive/MyDrive/zeros6.txt"
zeta_zeros = load_zetas(file_path)

# 📌 Sicherstellen, dass Daten vorhanden sind
if zeta_zeros is not None:
    print(f"✅ {len(zeta_zeros)} Nullstellen erfolgreich geladen.")

    # 📌 Optimierung starten
    optimal_beta, optimal_error = optimize_beta(zeta_zeros)

    # 📌 Ergebnisse ausgeben
    if optimal_beta is not None:
        print(f"🏆 Optimales Beta: {optimal_beta:.10f}")
        print(f"✅ Euler-Freese-Abweichung: {optimal_error:.10f}")

    # 📌 Plot der Beta-Abweichungen
    beta_values = np.linspace(0.45, 0.50, 50)
    errors = [euler_freese(zeta_zeros, beta) for beta in tqdm(beta_values)]

    plt.figure(figsize=(8, 5))
    plt.plot(beta_values, errors, "bo-", label="Euler-Freese-Abweichung")
    plt.axvline(optimal_beta, color="r", linestyle="dashed", label=f"Optimum: {optimal_beta:.5f}")
    plt.xlabel("Beta-Wert")
    plt.ylabel("Euler-Freese-Abweichung")
    plt.title("Optimierung von Beta für die Euler-Freese-Identität")
    plt.legend()
    plt.grid()
    plt.show()
else:
    print("❌ Keine Berechnung durchgeführt – Nullstellen ungültig.")
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install numpy scipy pandas matplotlib

# 📌 Importiere Bibliotheken
import pandas as pd
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt

# 📌 **Dateipfad für Nullstellen (bereinigte CSV!)**
csv_path = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"

# 📌 **Maximale Anzahl zu ladender Nullstellen & Eigenwerte**
N = 10000  # Kleinere Matrix für schnelleres Testen!
dx = 1 / N  # Optimierte Diskretisierung

# 📌 **Zeta-Nullstellen aus CSV laden & bereinigen**
try:
    zeta_csv = pd.read_csv(csv_path, header=None, dtype=str, nrows=N)
    zeta_csv = zeta_csv.iloc[:, 0].str.replace(",", "").astype(float)  # Entfernt Kommas & wandelt um
    zeta_zeros = zeta_csv.values
    print(f"✅ {N} Zeta-Nullstellen erfolgreich geladen!")

except Exception as e:
    print(f"❌ Fehler beim Laden der Datei: {e}")
    zeta_zeros = np.linspace(0, N, N)  # Fallback: Dummy-Daten

# 📌 **Potenzial-Funktion definieren**
def potential(n, A, β, C, D):
    return A * np.log(n + 1) + β / (n + 1) + C * np.sin(D * n)

# 📌 **Optimierte Parameter (aus vorherigen Analysen)**
A, β, C, D = 88.48780, -16352.98730, 0.5, 0.01

# 📌 **Hamilton-Operator definieren (Matrix H)**
diag = np.array([potential(n, A, β, C, D) for n in range(1, N+1)])
H = np.diag(diag) + np.diag(np.ones(N-1) * dx, k=1) + np.diag(np.ones(N-1) * dx, k=-1)

# 📌 **Eigenwerte & Eigenvektoren berechnen**
eigenvalues, eigenvectors = la.eigh(H)

# 📌 **Ergebnisse anzeigen**
print(f"🔢 Erste 10 Eigenwerte des Hamilton-Operators:\n{eigenvalues[:10]}")
print(f"🔍 Min/Max der Eigenwerte: {np.min(eigenvalues)}, {np.max(eigenvalues)}")

# 📌 **Fourier-Analyse der Eigenwerte**
fft_spectrum = np.fft.fft(eigenvalues)
frequencies = np.fft.fftfreq(len(eigenvalues))

# 📌 **Dominante Frequenzen extrahieren**
dominant_indices = np.argsort(np.abs(fft_spectrum))[-5:]
dominant_frequencies = frequencies[dominant_indices]
dominant_amplitudes = np.abs(fft_spectrum[dominant_indices])

print(f"🔥 Dominante Frequenzen: {np.round(dominant_frequencies, 4)}")
print(f"🔥 Korrespondierende Amplituden: {np.round(dominant_amplitudes, 4)}")

# 📌 **Korrelation der Eigenwerte mit den Zeta-Nullstellen berechnen**
correlation = np.corrcoef(eigenvalues, zeta_zeros)[0, 1]
print(f"📊 Korrelation der Eigenwerte mit Zeta-Nullstellen: {correlation:.5f}")

# 📌 **Plot: Eigenwerte vs. Index**
plt.figure(figsize=(7, 5))
plt.plot(eigenvalues, "bo-", markersize=3, label="Eigenwerte")
plt.xlabel("Index")
plt.ylabel("Eigenwerte")
plt.title("Spektrum des Hamilton-Operators")
plt.legend()
plt.grid()
plt.show()

# 📌 **Plot: Fourier-Spektrum der Eigenwerte**
plt.figure(figsize=(7, 5))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_spectrum)[:len(frequencies)//2], "r-")
plt.xlabel("Frequenz")
plt.ylabel("Spektrale Intensität")
plt.title("Fourier-Spektrum der Eigenwerte")
plt.grid()
plt.show()
```

```python
# 📌 Notwendige Bibliotheken installieren
!pip install pandas numpy matplotlib scipy gdown --quiet

# 📌 Bibliotheken importieren
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.fftpack as fft
import gdown

# 📌 **Google Drive Links der CSV-Dateien**
file_links = {
    "freese_function_cleaned.csv": "https://drive.google.com/uc?id=1ZL0uigQDakQomhD6bjKZPRjyae8x6v29",
    "freese_function_odlyzko_cleaned.csv": "https://drive.google.com/uc?id=1-6nkBCShsLsbMrR2uNJnUNZm4OvhmU_6",
    "freese_function_results.csv": "https://drive.google.com/uc?id=1-IIzV2Q5Qb-LnkRsznAwidKJR5Jqexdo",
    "freese_function_odlyzko.csv": "https://drive.google.com/uc?id=1-Oho0uYq7PPhTyyGq8e_B_FQLeMP61yy",
}

# 📌 **Chunk-weise Verarbeitung**
chunk_size = 100000  # 100k Zeilen pro Chunk
dataframes = {}

for name, link in file_links.items():
    output = f"/content/{name}"  # Speicherort für Colab
    gdown.download(link, output, quiet=False)

    try:
        # **Nur numerische Daten aus erster Spalte laden**
        df_list = []
        for chunk in pd.read_csv(output, header=None, skiprows=1, chunksize=chunk_size, dtype=str, on_bad_lines="skip"):
            chunk = chunk.iloc[:, 0]  # Erste Spalte wählen
            chunk = pd.to_numeric(chunk.str.replace(",", "."), errors="coerce")  # Strings in Float umwandeln
            chunk = chunk.dropna()  # Nicht-numerische Werte entfernen
            df_list.append(chunk)

        # **Finales DataFrame nur speichern, wenn nicht leer**
        if df_list:
            df = pd.concat(df_list, ignore_index=True)
            if not df.empty:
                dataframes[name] = df
            else:
                print(f"⚠️ {name} ist leer nach Bereinigung!")
        else:
            print(f"⚠️ {name} konnte nicht sinnvoll verarbeitet werden!")

    except Exception as e:
        print(f"⚠️ Fehler beim Einlesen von {name}: {e}")

# 📌 **Speicherverbrauch prüfen**
for name, df in dataframes.items():
    print(f"🔍 {name}: {df.shape} Einträge")
    print(df.head(), "\n")

# 📌 **Korrelationen berechnen (nur wenn gültige Daten vorhanden sind)**
correlations = {}
keys = list(dataframes.keys())

if len(keys) < 2:
    print("⚠️ Zu wenige gültige Datensätze für Korrelationen!")
else:
    for i in range(len(keys)):
        for j in range(i + 1, len(keys)):
            try:
                min_len = min(len(dataframes[keys[i]]), len(dataframes[keys[j]]))
                if min_len > 0:
                    corr = np.corrcoef(dataframes[keys[i]].iloc[:min_len], dataframes[keys[j]].iloc[:min_len])[0, 1]
                    correlations[f"{keys[i]} vs {keys[j]}"] = corr
            except Exception as e:
                print(f"⚠️ Fehler bei der Korrelation {keys[i]} vs {keys[j]}: {e}")

# 📌 **Korrelationsergebnisse ausgeben**
print("\n📊 **Korrelationen zwischen den Freese-Funktionen:**")
if correlations:
    for pair, corr in correlations.items():
        print(f"{pair}: {corr:.5f}")
else:
    print("⚠️ Keine gültigen Korrelationen berechnet!")

# 📌 **Fourier-Analyse der Freese-Funktionen (nur wenn Werte vorhanden sind)**
plt.figure(figsize=(12, 6))
for name, df in dataframes.items():
    if not df.empty:
        values = df.values[:min(len(df), 100000)]  # Maximale Anzahl an Daten begrenzen
        if len(values) > 0:
            freq_spectrum = np.abs(fft.fft(values))
            freqs = fft.fftfreq(len(values))

            plt.plot(freqs[:len(freqs)//2], freq_spectrum[:len(freq_spectrum)//2], label=name)
        else:
            print(f"⚠️ Keine Daten für FFT in {name}")

plt.xlabel("Frequenz")
plt.ylabel("Amplitude")
plt.title("📊 Fourier-Spektrum der Freese-Funktionen")
plt.legend()
plt.grid()
plt.show()
```
