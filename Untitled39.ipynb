```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Zielfunktion zur Minimierung der Abweichung von -1
def objective(beta):
    return np.abs(np.exp(beta * np.pi * 1j) + 1)

# Optimierung zur Bestimmung des besten Beta-Werts
result = opt.minimize_scalar(objective, bounds=(0, 1), method='bounded')

if result.success:
    beta_optimal = result.x
    exp_value = np.exp(beta_optimal * np.pi * 1j)
    deviation = np.abs(exp_value + 1)
else:
    beta_optimal = None
    exp_value = None
    deviation = None

# Vergleich mit Skalierungsrelationen
scale_1 = 1 / 129.4
scale_2 = 1 / 137
scale_3 = 9 / 200

diff_1 = np.abs(beta_optimal - scale_1)
diff_2 = np.abs(beta_optimal - scale_2)
diff_3 = np.abs(beta_optimal - scale_3)

# Ausgabe der Ergebnisse
print(f"Optimaler Beta-Wert: {beta_optimal:.15f}")
print(f"Wert von e^(beta * pi * i): {exp_value}")
print(f"Abweichung von -1: {deviation:.15e}")
print("\nVergleich der Beta-Abweichung mit Skalierungsrelationen:")
print(f"Abweichung von 1/129.4: {diff_1:.15f}")
print(f"Abweichung von 1/137: {diff_2:.15f}")
print(f"Abweichung von 9/200: {diff_3:.15f}")

# Visualisierung der Korrektur
fig, ax = plt.subplots(figsize=(6,6))
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
ax.add_artist(circle)
ax.axvline(x=-1, color='r', linestyle='dashed', label='Euler-Identität (-1)')

if exp_value is not None:
    ax.scatter(exp_value.real, exp_value.imag, color='blue', label=f'Beta-Korrektur ($\epsilon$ = {deviation:.3e})')

# Achsen & Labels
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel("Re($e^{\beta \pi i}$)")
ax.set_ylabel("Im($e^{\beta \pi i}$)")
ax.legend()
ax.set_title("Euler-Identität mit Beta-Korrektur")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definition der Siegel-Theta-Funktion
# Ansatz mit einer logarithmischen und trigonometrischen Komponente
def theta_siegel(t, A, beta, C, D, E, w, phi):
    return A * t**beta + C * np.log(t) + D * t**-1 + E * np.sin(w * np.log(t) + phi)

# Erzeugung von Testdaten für t (Indexwerte)
t = np.linspace(1, 1000, 1000)

# Berechnung der Fibonacci-Freese-Beta-Werte als Referenz
beta_werte = 0.5 + 0.02 * np.sin(0.1 * np.log(t))  # Näherung einer Doppel-Helix-Struktur

# Curve Fitting zur Anpassung der Siegel-Theta-Funktion an die Beta-Werte
popt, pcov = curve_fit(theta_siegel, t, beta_werte, p0=[1, 0.5, 0.01, 0.01, 0.02, 0.1, 0])

# Berechnete Werte der Theta-Funktion mit den optimalen Parametern
theta_fit = theta_siegel(t, *popt)

# Plot des Vergleichs
plt.figure(figsize=(10, 5))
plt.plot(t, beta_werte, label='Beta-Werte (FFF)', color='blue')
plt.plot(t, theta_fit, label='Theta-Siegel-Fit', linestyle='dashed', color='red')
plt.axhline(0.5, color='black', linestyle='dotted', label='Kritische Linie')
plt.legend()
plt.xlabel("Index t")
plt.ylabel("Wert")
plt.title("Vergleich: Fibonacci-Freese-Formel vs. Theta-Siegel-Funktion")
plt.grid()
plt.show()

# Ausgabe der optimalen Parameter
print("Optimale Parameter für die Siegel-Theta-Funktion:")
print(f"A = {popt[0]:.6f}, beta = {popt[1]:.6f}, C = {popt[2]:.6f}, D = {popt[3]:.6f}, ")
print(f"E = {popt[4]:.6f}, w = {popt[5]:.6f}, phi = {popt[6]:.6f}")
```

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

# Zielfunktion: Minimierung der Abweichung zu -1
def objective(beta):
    return np.abs(np.exp(beta * np.pi * 1j) + 1)

# Optimierung zur Bestimmung des besten Beta-Werts
result = opt.minimize_scalar(objective, bounds=(0, 1), method='bounded')

if result.success:
    beta_optimal = result.x
    exp_value = np.exp(beta_optimal * np.pi * 1j)
    deviation = np.abs(exp_value + 1)
else:
    beta_optimal = None
    exp_value = None
    deviation = None

# Skalierungsrelationen
scale_values = {
    "1 - 1/129.4": 1 - 1/129.4,
    "1 - 1/137": 1 - 1/137,
    "1 - 9/200": 1 - 9/200
}

# Berechnung der Abweichungen
deviation_scales = {k: np.abs(beta_optimal - v) for k, v in scale_values.items()}

# Ausgabe der Ergebnisse
print(f"Optimaler Beta-Wert: {beta_optimal:.15f}")
print(f"Wert von e^(beta * pi * i): {exp_value}")
print(f"Abweichung von -1: {deviation:.15e}")

print("\nVergleich der Beta-Abweichung mit Skalierungsrelationen:")
for key, value in deviation_scales.items():
    print(f"Abweichung von {key}: {value:.15f}")

# Visualisierung des Ergebnisses im Einheitskreis
fig, ax = plt.subplots(figsize=(6,6))
circle = plt.Circle((0, 0), 1, color='gray', fill=False, linestyle='dashed')
ax.add_artist(circle)
ax.axvline(x=-1, color='r', linestyle='dashed', label="Euler-Identität (-1)")

if exp_value is not None:
    ax.scatter(exp_value.real, exp_value.imag, color='blue',
               label=f"Beta-Korrektur ($\epsilon$ = {deviation:.3e})")

# Achsen & Labels (FIX: Matplotlib LaTeX-Probleme gelöst)
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel(r"Re$(e^{\beta \pi i})$")
ax.set_ylabel(r"Im$(e^{\beta \pi i})$")
ax.legend()
ax.set_title("Euler-Identität mit Beta-Korrektur")
plt.grid()
plt.show()

# Schwarzschild-Krümmung für Beta-Skala
N = np.arange(1, 1001)
beta_values = beta_optimal / (1 + np.log(N))  # Beispielhafte Beta-Skalierung
schwarzschild_fit = 1 - 2 * (-0.1456) / N  # Näherung der Schwarzschild-Krümmung

# Visualisierung der Beta-Skala vs. Schwarzschild-Krümmung
plt.figure(figsize=(8,4))
plt.scatter(N, beta_values, color='purple', label="Beta-Werte", alpha=0.6)
plt.plot(N, schwarzschild_fit, 'r--', label="Schwarzschild-Fit, M=-0.1456")
plt.xlabel("Index")
plt.ylabel("Beta-Wert")
plt.title("Euler-Gleichung vs. Schwarzschild-Krümmung")
plt.legend()
plt.grid()
```
