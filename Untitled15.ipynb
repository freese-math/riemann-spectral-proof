```python
import numpy as np
import matplotlib.pyplot as plt
import cupy as cp  # GPU-optimierte Library f√ºr Sparse-Matrizen
import cupyx.scipy.sparse.linalg as cupy_sparse
import torch
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import eigsh
from google.colab import drive
import time

# ‚úÖ Google Drive mounten
drive.mount('/content/drive')

# ‚úÖ Datei mit echten Nullstellen laden
file_path = "/content/drive/MyDrive/zeros6.txt"
print(f"üìÇ Lade Nullstellen aus {file_path}...")
nullstellen = np.loadtxt(file_path)
abstaende = np.diff(nullstellen)
print(f"‚úÖ {len(abstaende)} Nullstellen-Abst√§nde erfolgreich berechnet!")

# ‚úÖ GPU-Check
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"üíª Nutze GPU: {torch.cuda.get_device_name(0) if device == 'cuda' else 'Keine GPU verf√ºgbar'}")

# ‚úÖ Funktion: GPU-optimierte Sparse-GOE-Matrix mit Fallback auf CPU
def generate_sparse_goe_matrix(N, k=500):
    """Erstellt eine Sparse-GOE-Matrix und berechnet Eigenwerte mit GPU oder CPU"""
    print(f"‚ö° Berechne Sparse-Matrix f√ºr GOE mit N={N} (GPU wenn m√∂glich, sonst CPU)...")

    start_time = time.time()
    main_diag = np.random.randn(N)
    off_diag = np.random.randn(N-1) / np.sqrt(2)

    row_indices = np.arange(N-1)
    col_indices = row_indices + 1

    if N <= 50000 and device == "cuda":
        try:
            print("üöÄ Berechnung auf GPU...")
            # ‚úÖ Konvertiere alles in `cupy` bevor `concatenate()`
            main_diag_cp = cp.array(main_diag)
            off_diag_cp = cp.array(off_diag)
            row_indices_cp = cp.array(row_indices)
            col_indices_cp = cp.array(col_indices)

            H = cp.sparse.csr_matrix((cp.concatenate([off_diag_cp, off_diag_cp]),
                                      (cp.concatenate([row_indices_cp, col_indices_cp]),
                                       cp.concatenate([col_indices_cp, row_indices_cp]))),
                                      shape=(N, N))

            eigenvalues = cupy_sparse.eigsh(H, k=min(k, N-1), which='LM', return_eigenvectors=False)
            elapsed_time = time.time() - start_time
            print(f"‚úÖ GPU-Berechnung abgeschlossen in {elapsed_time:.2f} Sekunden")
            return cp.asnumpy(eigenvalues)
        except cp.cuda.memory.OutOfMemoryError:
            print("‚ö†Ô∏è GPU hat nicht genug Speicher. Umschalten auf CPU...")
            torch.cuda.empty_cache()  # GPU-Speicher freigeben

    # ‚úÖ CPU-Fallback
    print("üíª Berechnung auf CPU...")
    H = csr_matrix((np.concatenate([off_diag, off_diag]),
                    (np.concatenate([row_indices, col_indices]), np.concatenate([col_indices, row_indices]))),
                    shape=(N, N))

    eigenvalues = eigsh(H, k=min(k, N-1), which='LM', return_eigenvectors=False)
    elapsed_time = time.time() - start_time
    print(f"‚úÖ CPU-Berechnung abgeschlossen in {elapsed_time:.2f} Sekunden")

    return eigenvalues

# ‚úÖ Matrix-Analyse: Nutzt GPU wenn m√∂glich, sonst CPU
def analyze_random_matrix(N):
    print(f"üî¨ Erzeuge eine {N}x{N} GOE-Matrix und berechne die Eigenwerte...")
    eigenvalues = generate_sparse_goe_matrix(N, k=500)
    return eigenvalues

# ‚úÖ Testlauf f√ºr verschiedene N-Werte
goe_abstaende_20000 = analyze_random_matrix(20000)
goe_abstaende_50000 = analyze_random_matrix(50000)
goe_abstaende_100000 = analyze_random_matrix(100000)  # Sollte automatisch CPU w√§hlen

# ‚úÖ Vergleich der Abstandsverteilungen
plt.figure(figsize=(10, 6))
plt.hist(abstaende, bins=100, density=True, alpha=0.7, color='g', label="Riemann Nullstellen-Abst√§nde")
plt.hist(goe_abstaende_20000, bins=100, density=True, alpha=0.5, color='b', label="GOE Eigenwert-Abst√§nde (N=20.000)")
plt.hist(goe_abstaende_50000, bins=100, density=True, alpha=0.5, color='r', label="GOE Eigenwert-Abst√§nde (N=50.000)")
plt.xlabel("Abst√§nde")
plt.ylabel("Dichte")
plt.title("Vergleich: Riemann Nullstellen vs. GOE Level Spacings")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ‚úÖ Lorentz-Transformation definieren
def lorentz_transform(x, y, z, t, beta):
    """
    F√ºhrt eine Lorentz-Transformation entlang der z-Achse durch.
    beta: Geschwindigkeit als Bruchteil der Lichtgeschwindigkeit c (0 <= beta < 1)
    """
    gamma = 1 / np.sqrt(1 - beta**2)  # Lorentz-Faktor
    z_prime = gamma * (z - beta * t)
    t_prime = gamma * (t - beta * z)
    return x, y, z_prime, t_prime

# ‚úÖ Beispieldaten: Nullstellen als Punkte im Raum
N = 500
t_values = np.linspace(0, 50, N)  # Zeitwerte f√ºr Lichtkegel-Struktur
x_values = np.sin(t_values) * 10  # Simulierte x-Koordinaten (analog Nullstellenverteilung)
y_values = np.cos(t_values) * 10  # Simulierte y-Koordinaten
z_values = t_values  # Initiale z-Koordinaten entlang der z-Achse

# ‚úÖ Lorentz-Transformation anwenden
beta = 0.9  # Relativistische Geschwindigkeit (90% der Lichtgeschwindigkeit)
x_trans, y_trans, z_trans, t_trans = lorentz_transform(x_values, y_values, z_values, t_values, beta)

# ‚úÖ 3D-Visualisierung des transformierten Lichtkegels
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Vorher: Originale Nullstellen
ax.scatter(x_values, y_values, z_values, color='blue', alpha=0.5, label="Originale Nullstellen")

# Nachher: Lorentz-transformierte Nullstellen
ax.scatter(x_trans, y_trans, z_trans, color='red', alpha=0.8, label="Transformierte Nullstellen")

# Achsentitel setzen
ax.set_xlabel("X-Koordinate")
ax.set_ylabel("Y-Koordinate")
ax.set_zlabel("Z-Koordinate (Zeit)")

ax.set_title("Lorentz-Transformation der Nullstellen")
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

# Funktion zur Erzeugung einer GOE/GUE-Matrix und Berechnung der Eigenwerte
def generate_goe_matrix(N):
    """ Erzeugt eine symmetrische GOE-Matrix der Gr√∂√üe NxN """
    A = np.random.randn(N, N)
    return (A + A.T) / np.sqrt(2 * N)

def generate_gue_matrix(N):
    """ Erzeugt eine hermitesche GUE-Matrix der Gr√∂√üe NxN """
    A = np.random.randn(N, N) + 1j * np.random.randn(N, N)
    return (A + A.T.conj()) / np.sqrt(2 * N)

def eigenvalue_spacing(N, matrix_type="GOE"):
    """ Berechnet die Eigenwerte und Eigenwert-Abst√§nde f√ºr eine zuf√§llige Matrix """
    if matrix_type == "GOE":
        M = generate_goe_matrix(N)
    elif matrix_type == "GUE":
        M = generate_gue_matrix(N)
    else:
        raise ValueError("matrix_type muss 'GOE' oder 'GUE' sein.")

    eigenvalues = np.linalg.eigh(M)[0]
    eigenvalues.sort()
    spacings = np.diff(eigenvalues)  # Abst√§nde zwischen Eigenwerten
    return spacings

# Berechnungen f√ºr verschiedene Matrixgr√∂√üen
N_small = 20000
N_large = 50000

print("üî¨ Berechne Eigenwert-Abst√§nde f√ºr GOE (N=20000)...")
goe_abstaende_20000 = eigenvalue_spacing(N_small, "GOE")

print("üî¨ Berechne Eigenwert-Abst√§nde f√ºr GUE (N=20000)...")
gue_abstaende_20000 = eigenvalue_spacing(N_small, "GUE")

print("üî¨ Berechne Eigenwert-Abst√§nde f√ºr GOE (N=50000)...")
goe_abstaende_50000 = eigenvalue_spacing(N_large, "GOE")

print("üî¨ Berechne Eigenwert-Abst√§nde f√ºr GUE (N=50000)...")
gue_abstaende_50000 = eigenvalue_spacing(N_large, "GUE")

# Histogramm der Eigenwert-Abst√§nde
plt.figure(figsize=(10, 6))
plt.hist(goe_abstaende_20000, bins=100, alpha=0.6, label="GOE (N=20.000)", color='blue', density=True)
plt.hist(gue_abstaende_20000, bins=100, alpha=0.6, label="GUE (N=20.000)", color='red', density=True)
plt.hist(goe_abstaende_50000, bins=100, alpha=0.6, label="GOE (N=50.000)", color='green', density=True)
plt.xlabel("Abst√§nde")
plt.ylabel("Dichte")
plt.legend()
plt.title("Vergleich: Riemann Nullstellen vs. GOE Level Spacings")
plt.show()
```
