```python
# âœ… Installiere benÃ¶tigte Bibliotheken
!pip install numpy scipy matplotlib gudhi

import numpy as np
import scipy.optimize as opt
import scipy.fftpack as fft
import matplotlib.pyplot as plt
import gudhi as gd

# âœ… Reduziere Speicherverbrauch durch weniger Datenpunkte
max_data_size = 500000  # Statt 2 Mio. nur 500.000 Werte laden

# âœ… Lade die echten Odlyzko-Nullstellen (2 Mio.)
file_path = "/content/drive/MyDrive/zeros6.txt"
zeros = np.loadtxt(file_path)[:max_data_size]  # Lade nur die ersten 500k

# âœ… Berechne Nullstellen-AbstÃ¤nde
abstaende = np.diff(zeros)

# âœ… Definiere die erweiterte Freese-Funktion mit Korrekturtermen
def freese_function(N, A, beta, B, C):
    return A * (N ** beta) * (np.log(N) ** B) + C

# âœ… Werte fÃ¼r N erzeugen
N_values = np.arange(1, len(abstaende) + 1)
y_values = np.cumsum(abstaende)

# âœ… Nichtlineare Anpassung der FF-Funktion
popt, _ = opt.curve_fit(freese_function, N_values, y_values, p0=[2.0, 0.91, 0.1, 0.0])
A_opt, beta_opt, B_opt, C_opt = popt

# âœ… Fourier-Analyse auf reduzierten Daten (Speicheroptimierung)
fft_data_size = 100000  # Nur 100k Werte fÃ¼r FFT
fft_spectrum = np.abs(fft.fft(abstaende[:fft_data_size]))
freqs = fft.fftfreq(fft_data_size)

# âœ… Signifikante Frequenzen extrahieren
significant_freqs = freqs[np.argsort(fft_spectrum)[-2:]]  # Top-2 Frequenzen

# âœ… Approximation der fraktalen Dimension (Box-Counting)
def box_counting(data, epsilon):
    counts = []
    for eps in epsilon:
        counts.append(np.unique(np.floor(data / eps)).size)
    coeffs = np.polyfit(np.log(epsilon), np.log(counts), 1)
    return -coeffs[0]  # Fraktale Dimension

epsilons = np.logspace(-2, 1, 10)  # Werte fÃ¼r Box-Counting
fractal_dimension = box_counting(abstaende, epsilons)

# âœ… Ergebnisse ausgeben
print(f"ðŸ“Œ Optimierte Parameter fÃ¼r FF:")
print(f"ðŸ”¹ A = {A_opt}")
print(f"ðŸ”¹ Î² = {beta_opt}")
print(f"ðŸ”¹ B = {B_opt}")
print(f"ðŸ”¹ C = {C_opt}\n")

print(f"ðŸ“Œ Signifikante Frequenzen in der Fourier-Analyse: {significant_freqs}")
print(f"ðŸ“Œ Fraktale Dimension (Box-Counting-Methode): {fractal_dimension}")

# âœ… Ergebnisse visualisieren
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Echte FF-Daten", color="blue")
plt.plot(N_values, freese_function(N_values, A_opt, beta_opt, B_opt, C_opt),
         linestyle="dashed", color="red", label="Optimierte FF-Funktion")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("N")
plt.ylabel("Î£ Nullstellen-AbstÃ¤nde")
plt.legend()
plt.title("Vergleich der Freese-Funktion mit der Riemannschen NÃ¤herung")
plt.show()

# âœ… Speichern der Ergebnisse
output_file = "/content/drive/MyDrive/freese_function_odlyzko.csv"
np.savetxt(output_file, np.column_stack([N_values, y_values]), delimiter=",",
           header="N,Echte_Summe_Nullstellen_AbstÃ¤nde", comments="")
print(f"âœ… Datei gespeichert: {output_file}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from scipy.sparse import diags

# ðŸ“Œ **Effektives Potential aus Operatorstruktur ableiten**
def effektives_potential(x, a=1.0, b=0.5):
    return a * x**b

# ðŸ“Œ **Erzeuge Werte fÃ¼r das effektive Potential**
x_values = np.linspace(1, 500, 500)
V_values = effektives_potential(x_values)

# ðŸ“Œ **Plot des effektiven Potentials**
plt.figure(figsize=(8,5))
plt.plot(x_values, V_values, label=r"Effektives Potential $V(x)$")
plt.xlabel("Index x")
plt.ylabel("V(x)")
plt.title(r"Abgeleitetes effektives Potential aus $\hat{H}$")
plt.legend()
plt.show()

# ðŸ“Œ **Vergleich mit Zufallsmatrizen (GOE)**
def random_matrix(n):
    """Erzeugt eine Zufallsmatrix aus der Gaussian Orthogonal Ensemble (GOE)."""
    M = np.random.normal(0, 1, (n, n))
    return (M + M.T) / 2  # Symmetrisieren

# ðŸ“Œ **MatrixgrÃ¶ÃŸe fÃ¼r GOE**
n = 200

# ðŸ“Œ **Erstelle Operator-Matrix basierend auf NÃ¤herungen**
H_operator = diags([V_values[:n]], [0]).toarray()

# ðŸ“Œ **Berechne Eigenwerte**
eigvals_operator = eigh(H_operator, eigvals_only=True)
eigvals_goe = eigh(random_matrix(n), eigvals_only=True)

# ðŸ“Œ **Histogramm der Eigenwerte**
plt.figure(figsize=(10,5))
plt.hist(eigvals_operator, bins=50, density=True, alpha=0.7, color="green", label="Operator Eigenwerte")
plt.hist(eigvals_goe, bins=50, density=True, alpha=0.5, color="blue", label="GOE-Eigenwerte")
plt.axvline(0, color="red", linestyle="dashed", linewidth=2, label="Perfekte Symmetrie")
plt.xlabel("Eigenwert")
plt.ylabel("HÃ¤ufigkeit")
plt.title("Vergleich mit Zufallsmatrizen & Quantenchaos")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg

# ðŸ“Œ **Operator-Definition fÃ¼r \hat{H}**
N = 500  # GrÃ¶ÃŸe der Matrix
A = np.random.normal(0, 1, (N, N))  # ZufÃ¤llige Matrix zur Simulation
H = (A + A.T) / 2  # Hermitescher Operator

# ðŸ“Œ **Eigenwerte berechnen**
eigenvalues = np.linalg.eigvalsh(H)

# ðŸ“Œ **Vergleich mit Zufallsmatrizen (GOE)**
GOE = np.random.normal(0, 1, (N, N))
GOE = (GOE + GOE.T) / 2  # Symmetrische GOE-Matrix
GOE_eigenvalues = np.linalg.eigvalsh(GOE)

# ðŸ“Œ **Histogramm der Eigenwerte**
plt.figure(figsize=(10, 5))
plt.hist(eigenvalues, bins=50, density=True, alpha=0.7, color='g', label="Operator Eigenwerte")
plt.hist(GOE_eigenvalues, bins=50, density=True, alpha=0.5, color='b', label="GOE-Eigenwerte")
plt.axvline(x=0, color='r', linestyle='dashed', linewidth=2, label="Perfekte Symmetrie")
plt.legend()
plt.title("Vergleich mit Zufallsmatrizen & Quantenchaos")
plt.xlabel("Eigenwert")
plt.ylabel("HÃ¤ufigkeit")
plt.show()

# ðŸ“Œ **Effektives Potential ableiten**
x = np.arange(1, N+1)
V_x = np.sqrt(x)  # Effektives Potential V(x) âˆ sqrt(x)

plt.figure(figsize=(8, 5))
plt.plot(x, V_x, label="Effektives Potential V(x)")
plt.xlabel("Index x")
plt.ylabel("V(x)")
plt.title("Abgeleitetes effektives Potential aus \hat{H}")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
from scipy.optimize import curve_fit

# ðŸ“Œ **SchrÃ¶dinger-Operator und Zufallsmatrizen-Analyse**
# Definiere die Matrix-Form von \hat{H} fÃ¼r die Eigenwertanalyse

def generate_operator_matrix(N, alpha=1.0, beta=0.5):
    """Erzeugt eine Operator-Matrix Ã¤hnlich einem SchrÃ¶dinger-Hamiltonian"""
    H = np.zeros((N, N))
    for i in range(N):
        H[i, i] = alpha * (i+1)**beta  # Diagonalelemente als "Potential"
        if i < N - 1:
            H[i, i+1] = H[i+1, i] = -1  # Kopplung benachbarter ZustÃ¤nde
    return H

# ðŸ“Œ **Berechne Eigenwerte fÃ¼r Operator H**
N = 500
H = generate_operator_matrix(N)
eigenvalues, eigenvectors = eigh(H)

# ðŸ“Œ **Vergleich mit Zufallsmatrizen GOE (Gaussian Orthogonal Ensemble)**
random_matrix = np.random.randn(N, N)
random_matrix = (random_matrix + random_matrix.T) / 2  # Symmetrische GOE-Matrix
rand_eigenvalues, _ = eigh(random_matrix)

# ðŸ“Œ **Histogramm der Eigenwerte vergleichen**
plt.figure(figsize=(10, 5))
plt.hist(eigenvalues, bins=50, density=True, alpha=0.6, color='g', label="Operator Eigenwerte")
plt.hist(rand_eigenvalues, bins=50, density=True, alpha=0.4, color='b', label="GOE-Eigenwerte")
plt.axvline(0, color='r', linestyle='dashed', linewidth=2, label="Perfekte Symmetrie")
plt.legend()
plt.title("Vergleich mit Zufallsmatrizen & Quantenchaos")
plt.xlabel("Eigenwert")
plt.ylabel("HÃ¤ufigkeit")
plt.show()

# ðŸ“Œ **Effektives Potential V(x) aus der Operator-Diagonalen ableiten**
x_vals = np.arange(1, N+1)
V_vals = np.diagonal(H)  # Die Diagonale der Matrix entspricht dem effektiven Potential

plt.figure(figsize=(8, 5))
plt.plot(x_vals, V_vals, label="Effektives Potential V(x)")
plt.xlabel("Index x")
plt.ylabel("V(x)")
plt.title("Abgeleitetes effektives Potential aus \hat{H}")
plt.legend()
plt.show()

# ðŸ“Œ **Fit fÃ¼r effektives Potential durchfÃ¼hren**
def potential_model(x, a, b):
    return a * x**b

params, _ = curve_fit(potential_model, x_vals, V_vals)
a_fit, b_fit = params
print(f"Effektives Potential angenÃ¤hert durch: V(x) = {a_fit:.4f} * x^{b_fit:.4f}")

# ðŸ“Œ **SchrÃ¶dinger-Gleichung mit diesem Potential simulieren (numerische LÃ¶sung folgt spÃ¤ter)**
```

```python
# âœ… Notwendige Pakete installieren
!pip install numpy scipy matplotlib gudhi

# âœ… Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fftpack import fft
import gudhi as gd
import pandas as pd

# âœ… Pfad zu den Original-Odlyzko-Nullstellen (Google Drive)
file_path = "/content/drive/MyDrive/zeros6.txt"

# ðŸ“Œ **Funktion zum Laden der Nullstellen**
def load_nullstellen(file_path):
    """LÃ¤dt Nullstellen aus einer Textdatei"""
    with open(file_path, "r") as f:
        return np.array([float(line.strip()) for line in f])

# âœ… Nullstellen laden
zeros = load_nullstellen(file_path)
print(f"âœ… {len(zeros)} Nullstellen erfolgreich geladen!")

# ðŸ“Œ **Berechnung der Nullstellen-AbstÃ¤nde**
abstaende = np.diff(zeros)  # Differenz der Nullstellen
N_values = np.arange(1, len(abstaende) + 1)  # Korrekte LÃ¤nge!
y_values = np.cumsum(abstaende)  # Kumulative Summe der AbstÃ¤nde

# ðŸ“Œ **Freese-Funktion (FF)**
def FF(N, A, beta, C, D, w, phi):
    return A * N**beta + C + D * np.sin(w * N + phi)

# ðŸ“Œ **Curve-Fitting zur Optimierung der FF**
popt, _ = curve_fit(FF, N_values, y_values, p0=[2.0, 0.91, 1000, 0.01, 0.08, -9000])
A_opt, beta_opt, C_opt, D_opt, w_opt, phi_opt = popt

print("\nðŸ“Œ **Optimierte Parameter fÃ¼r FF:**")
print(f"ðŸ”¹ A = {A_opt:.6f}")
print(f"ðŸ”¹ Î² = {beta_opt:.6f}")
print(f"ðŸ”¹ C = {C_opt:.6f}")
print(f"ðŸ”¹ D = {D_opt:.6f}")
print(f"ðŸ”¹ w = {w_opt:.6f}")
print(f"ðŸ”¹ Ï† = {phi_opt:.6f}")

# ðŸ“Œ **Fehlermetriken berechnen**
fehler = y_values - FF(N_values, *popt)
rms_fehler = np.sqrt(np.mean(fehler**2))

print(f"\nðŸ“Š **Fehlerstatistik:**")
print(f"ðŸ”¹ RMS-Fehler: {rms_fehler:.6f}")
print(f"ðŸ”¹ Maximaler Fehler: {np.max(fehler):.6f}")
print(f"ðŸ”¹ Minimaler Fehler: {np.min(fehler):.6f}")

# ðŸ“Œ **Fourier-Analyse der Fehler**
fft_fehler = np.abs(fft(fehler))[:len(fehler)//2]
frequenzen = np.fft.fftfreq(len(fehler), d=1)[:len(fehler)//2]

# ðŸ“Œ **Signifikante Frequenzen extrahieren**
signifikante_freqs = frequenzen[np.argsort(fft_fehler)[-5:]]
print(f"\nðŸ“Œ **Signifikante Frequenzen in der Fourier-Analyse:** {signifikante_freqs}")

# ðŸ“Œ **Betti-Zahlen berechnen (Topologische Analyse)**
data_2D = np.column_stack((np.real(zeros[:1000]), np.imag(zeros[:1000])))
rips = gd.RipsComplex(points=data_2D, max_edge_length=2)
simplex_tree = rips.create_simplex_tree(max_dimension=2)
simplex_tree.compute_persistence()  # Wichtig, bevor `betti_numbers()` aufgerufen wird!
betti_numbers = simplex_tree.betti_numbers()

print(f"\nðŸ“Œ **Betti-Zahlen fÃ¼r die Nullstellenverteilung:** {betti_numbers}")

# ðŸ“Œ **Ergebnisse als CSV speichern**
results = pd.DataFrame({
    "A": [A_opt],
    "Î²": [beta_opt],
    "C": [C_opt],
    "D": [D_opt],
    "w": [w_opt],
    "Ï†": [phi_opt],
    "RMS-Fehler": [rms_fehler],
    "Betti-Zahlen": [betti_numbers]
})
csv_path = "/content/drive/MyDrive/freese_function_results.csv"
results.to_csv(csv_path, index=False)
print(f"\nâœ… Datei gespeichert: {csv_path}")

# ðŸ“Œ **Plotte kumulierte AbstÃ¤nde und FF-Fit**
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Kumulierte AbstÃ¤nde", color="blue")
plt.plot(N_values, FF(N_values, *popt), '--', label="FF-Fit", color="red")
plt.xlabel("N")
plt.ylabel("Î£ AbstÃ¤nde")
plt.title("FF-Fit der kumulierten Nullstellen-AbstÃ¤nde")
plt.legend()
plt.grid()
plt.show()
```

```python
# âœ… Bibliotheken importieren
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
import scipy.fftpack as fft
import scipy.stats as stats
import sympy as sp
import gudhi as gd
from scipy.signal import find_peaks
from scipy.stats import pearsonr

# âœ… Google Drive mounten
from google.colab import drive
drive.mount('/content/drive')

# âœ… Dateipfad fÃ¼r Odlyzko-Nullstellen
pfad_nullstellen = "/content/drive/MyDrive/zeros6.txt"

# âœ… Daten laden
def load_numbers(file_path):
    with open(file_path, 'r') as f:
        return np.array([float(line.strip()) for line in f])

zeros = load_numbers(pfad_nullstellen)
N_total = len(zeros)
print(f"âœ… {N_total} Nullstellen erfolgreich geladen!")

# âœ… Berechnung der Nullstellen-AbstÃ¤nde
abstaende = np.diff(zeros)

# âœ… Definition der Freese-Funktion (FF)
def FF(N, A, beta, C, D, w, phi):
    return A * N**beta + C * np.log(N) + D * np.sin(w*N + phi)

# âœ… Curve Fitting zur Optimierung der FF-Parameter
N_values = np.arange(1, N_total)
y_values = np.cumsum(abstaende)

popt, _ = opt.curve_fit(FF, N_values, y_values, p0=[2.0, 0.91, 1000, 0.01, 0.08, -9000])
A_opt, beta_opt, C_opt, D_opt, w_opt, phi_opt = popt

# âœ… Ausgabe der optimierten Parameter
print(f"ðŸ“Œ Optimierte Parameter fÃ¼r FF:")
print(f"ðŸ”¹ A = {A_opt}")
print(f"ðŸ”¹ Î² = {beta_opt}")
print(f"ðŸ”¹ C = {C_opt}")
print(f"ðŸ”¹ D = {D_opt}")
print(f"ðŸ”¹ w = {w_opt}")
print(f"ðŸ”¹ Ï† = {phi_opt}")

# âœ… Fehlerberechnung
fehler = y_values - FF(N_values, A_opt, beta_opt, C_opt, D_opt, w_opt, phi_opt)

# âœ… Fourier-Analyse der Fehler
fft_fehler = np.abs(fft.fft(fehler))[:len(fehler)//2]
frequenzen = np.fft.fftfreq(len(fehler), d=1)[:len(fehler)//2]
peaks, _ = find_peaks(fft_fehler, height=np.max(fft_fehler) * 0.1)
signifikante_freqs = frequenzen[peaks]

print(f"\nðŸ“Œ Signifikante Frequenzen in der Fourier-Analyse: {signifikante_freqs}")

# âœ… Fraktale Dimension (Box-Counting-Methode)
def fractal_dimension(data, scale_range=(1, 100), num_scales=10):
    scales = np.logspace(*np.log10(scale_range), num=num_scales)
    counts = np.zeros_like(scales)

    for i, s in enumerate(scales):
        grid = np.floor(data / s)
        counts[i] = len(np.unique(grid))

    coeffs = np.polyfit(np.log(scales), np.log(counts), 1)
    return -coeffs[0]

D_f = fractal_dimension(y_values)
print(f"ðŸ“Œ Fraktale Dimension (Box-Counting-Methode): {D_f}")

# âœ… Topologische Analyse (Betti-Zahlen)
data_2D = np.column_stack((np.real(zeros[:1000]), np.imag(zeros[:1000])))
rips = gd.RipsComplex(points=data_2D, max_edge_length=2)
simplex_tree = rips.create_simplex_tree(max_dimension=2)
simplex_tree.compute_persistence()
betti_numbers = simplex_tree.betti_numbers()
print(f"ðŸ“Œ Betti-Zahlen fÃ¼r die Nullstellenverteilung: {betti_numbers}")

# âœ… Operator-Darstellung der FF
N = sp.Symbol('N', real=True, positive=True)
A, beta, C, D, w, phi = sp.symbols('A beta C D w phi', real=True)

FF_operator = A * N**beta + C * sp.log(N) + D * sp.sin(w*N + phi)
dFF_dN = sp.diff(FF_operator, N)  # Erste Ableitung
d2FF_dN2 = sp.diff(dFF_dN, N)     # Zweite Ableitung

print("\nðŸ“Œ Operator-Darstellung der Freese-Funktion:")
print("ðŸ”¹ FF(N) =", FF_operator)
print("ðŸ”¹ d/dN FF(N) =", dFF_dN)
print("ðŸ”¹ dÂ²/dNÂ² FF(N) =", d2FF_dN2)

# âœ… Ergebnisse speichern
ergebnisse = np.column_stack((N_values, y_values, fehler))
np.savetxt("/content/drive/MyDrive/freese_function_results.csv", ergebnisse, delimiter=",", header="N, y_values, Fehler", comments='')

print("âœ… Datei gespeichert: /content/drive/MyDrive/freese_function_results.csv")

# âœ… Visualisierung
plt.figure(figsize=(10,5))
plt.plot(N_values, y_values, label="Kumulative AbstÃ¤nde", color="blue")
plt.plot(N_values, FF(N_values, A_opt, beta_opt, C_opt, D_opt, w_opt, phi_opt), '--', label="FF-Fit", color="red")
plt.xlabel("N")
plt.ylabel("Î£ AbstÃ¤nde")
plt.title("Freese-Funktion Fit")
plt.legend()
plt.grid()
plt.show()
```
