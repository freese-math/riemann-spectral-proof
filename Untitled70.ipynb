```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.fft import fft, fftfreq

# === 1. Daten laden ===
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
df = pd.read_csv(beta_path, skiprows=1, header=None)
df = df.apply(pd.to_numeric, errors='coerce').dropna()
beta = df.iloc[:, 0].astype(float).values  # Annahme: β(n) in der zweiten Spalte

# === 2. Datenbereich einschränken ===
window_size = 20000
beta_window = beta[:window_size]
n = np.arange(1, window_size + 1)

# === 3. Mittelwert entfernen (Zentrierung) ===
beta_centered = beta_window - np.mean(beta_window)

# === 4. Dominante Frequenz mittels Fourier-Analyse bestimmen ===
N = len(n)
yf = fft(beta_centered)
xf = fftfreq(N, 1)[:N//2]
dominant_freq = xf[np.argmax(2.0/N * np.abs(yf[0:N//2]))]

# === 5. Modellfunktion definieren ===
def beta_model(n, a, b, f, phi):
    return a / np.log(n) + b * np.cos(2 * np.pi * f * n + phi)

# === 6. Startwerte und Grenzen festlegen ===
p0 = [1e5, 1e5, dominant_freq, 0]  # Startwerte basierend auf Fourier-Analyse
bounds = ([0, 0, 0, -np.pi], [np.inf, np.inf, np.inf, np.pi])  # Parametergrenzen

# === 7. Curve Fit durchführen ===
params, pcov = curve_fit(beta_model, n, beta_centered, p0=p0, bounds=bounds, method='trf', maxfev=10000)
a, b, f_fit, phi = params

# === 8. Modellwerte berechnen ===
beta_fit = beta_model(n, a, b, f_fit, phi)

# === 9. Plot: Vergleich zwischen Daten und Fit ===
plt.figure(figsize=(12, 5))
plt.plot(n, beta_centered, label=r"$\beta(n)$ – zentriert", color="blue")
plt.plot(n, beta_fit, label=fr"Fit: $a/\log(n) + b \cdot \cos(2\pi \cdot {f_fit:.5f} \cdot n + \varphi)$", color="red")
plt.title("Fit der zentrierten Beta-Skala mit optimierten Parametern")
plt.xlabel("n")
plt.ylabel(r"$\beta(n)$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === 10. Fit-Parameter und Unsicherheiten ausgeben ===
perr = np.sqrt(np.diag(pcov))
print("✅ Gefittete Parameter:")
print(f"  a = {a:.4f} ± {perr[0]:.4f}")
print(f"  b = {b:.4f} ± {perr[1]:.4f}")
print(f"  f = {f_fit:.6f} ± {perr[2]:.6f}")
print(f"  φ = {phi:.4f} ± {perr[3]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Parameter ---
N = 1000
n = np.arange(N)
frequenzen = [0.01, 0.015, 0.02]
amplituden = [50, 30, 20]
phasen = [0, np.pi / 3, np.pi / 4]
rauschpegel = 1.0
threshold = 100  # FFT-Schwellenwert

# --- Signal erzeugen ---
def generiere_signal(n, frequenzen, amplituden, phasen):
    return sum(
        A * np.cos(2 * np.pi * f * n + phi)
        for A, f, phi in zip(amplituden, frequenzen, phasen)
    )

beta_n_clean = generiere_signal(n, frequenzen, amplituden, phasen)
np.random.seed(0)
noise = rauschpegel * np.random.normal(size=N)
beta_n_noisy = beta_n_clean + noise

# --- FFT und Spektralanalyse ---
def plot_fft(signal, title, threshold=None):
    fft_vals = np.fft.fft(signal)
    freqs = np.fft.fftfreq(len(signal), d=1)
    spectrum = np.abs(fft_vals[:N // 2])
    pos_freqs = freqs[:N // 2]

    plt.figure(figsize=(12, 4))
    plt.plot(pos_freqs, spectrum, label='Spektrum')
    if threshold:
        plt.axhline(threshold, color='r', linestyle='--', label='Schwelle')
    plt.title(title)
    plt.xlabel("Frequenz")
    plt.ylabel("Spektrale Stärke")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    return fft_vals

# --- Plots ---
fft_clean = plot_fft(beta_n_clean, "FFT von βₙ ohne Rauschen", threshold)
fft_noisy = plot_fft(beta_n_noisy, "FFT von βₙ mit Rauschen", threshold)

# --- Frequenzfilterung & Rekonstruktion ---
def rekonstruieren(fft_vals, threshold):
    filtered_fft = np.where(np.abs(fft_vals) > threshold, fft_vals, 0)
    return np.fft.ifft(filtered_fft).real, filtered_fft

recon, filtered_fft = rekonstruieren(fft_clean, threshold)

# --- Vergleich ---
plt.figure(figsize=(12, 4))
plt.plot(n, beta_n_clean, label="Original")
plt.plot(n, recon, '--', label="Rekonstruiert")
plt.title("Rekonstruktion von βₙ aus Spektrum")
plt.xlabel("n")
plt.ylabel(r"$\beta_n$")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Fehleranalyse ---
max_error = np.max(np.abs(beta_n_clean - recon))
print(f"Maximaler Rekonstruktionsfehler: {max_error:.2e}")

# --- Optional: Frequenzmarken der stärksten Peaks anzeigen ---
spectrum_mag = np.abs(filtered_fft[:N // 2])
top_freq_indices = np.argsort(spectrum_mag)[-5:][::-1]
top_freqs = np.fft.fftfreq(N, d=1)[:N // 2][top_freq_indices]
top_values = spectrum_mag[top_freq_indices]

print("\nDominante Frequenzkomponenten:")
for i, (f, val) in enumerate(zip(top_freqs, top_values), 1):
    print(f"{i}. Frequenz: {f:.4f}, Stärke: {val:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.interpolate import interp1d
from scipy.stats import pearsonr

# --- Hilfsfunktionen ---
def einlesen_beta_skala(dateipfad):
    """Liest eine Beta-Skala ein und gibt die Beta-Werte zurück."""
    df = pd.read_csv(dateipfad, header=None)
    if df.shape[1] == 2:
        beta_werte = df.iloc[:, 1].astype(float).values
    else:
        beta_werte = df.iloc[:, 0].astype(float).values
    return beta_werte

def interpolieren(skala, ziel_länge):
    """Interpoliert eine Skala auf die angegebene Ziellänge."""
    x_alt = np.linspace(0, 1, num=len(skala))
    x_neu = np.linspace(0, 1, num=ziel_länge)
    interpolator = interp1d(x_alt, skala, kind='linear')
    return interpolator(x_neu)

def berechne_differenzen(werte):
    """Berechnet die Differenzen zwischen aufeinanderfolgenden Werten."""
    return np.diff(werte)

def normalisieren(daten):
    """Normalisiert die Daten auf Mittelwert 0 und Standardabweichung 1."""
    return (daten - np.mean(daten)) / np.std(daten)

def plot_fft(signal, title):
    """Erstellt ein FFT-Plot für das gegebene Signal."""
    n = len(signal)
    freq = fftfreq(n, d=1)[:n // 2]
    spectrum = np.abs(fft(signal))[:n // 2]

    plt.figure(figsize=(10, 4))
    plt.plot(freq, spectrum)
    plt.title(f"FFT: {title}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# --- Dateipfade ---
pfad_nullstellen = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
pfade_beta_skalen = [
    "/content/drive/MyDrive/beta_scale_operator_cleaned.csv",
    "/content/drive/MyDrive/beta_skala_praezision_cleaned.csv",
    "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv",
    "/content/drive/MyDrive/beta_gpu_stabilisiert.csv",
    "/content/drive/MyDrive/beta_gpu_interpoliert.csv",
    "/content/drive/MyDrive/beta_skala_kombi_repaired.csv"
]

# --- Einlesen der Nullstellen ---
nullstellen = pd.read_csv(pfad_nullstellen, header=None).squeeze("columns").astype(float).values
nullstellen_diffs = berechne_differenzen(nullstellen)
nullstellen_diffs_norm = normalisieren(nullstellen_diffs)

# --- Verarbeitung jeder Beta-Skala ---
for pfad in pfade_beta_skalen:
    beta_werte = einlesen_beta_skala(pfad)

    # Interpolation, falls nötig
    if len(beta_werte) != len(nullstellen):
        beta_werte = interpolieren(beta_werte, len(nullstellen))

    beta_diffs = berechne_differenzen(beta_werte)
    beta_diffs_norm = normalisieren(beta_diffs)

    # Berechnung der Pearson-Korrelation
    korrelation, _ = pearsonr(nullstellen_diffs_norm, beta_diffs_norm)
    print(f"Korrelation zwischen Nullstellen-Differenzen und Beta-Differenzen ({pfad}): {korrelation:.4f}")

    # FFT-Analyse
    plot_fft(beta_diffs_norm, f"Beta-Differenzen (normiert) - {pfad}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten sicher laden ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"

# Versuche nur numerische Werte zu laden
df = pd.read_csv(data_path, header=None, dtype=str)
df = df.apply(pd.to_numeric, errors='coerce')  # ungültige Werte zu NaN
df = df.dropna()  # NaN-Zeilen entfernen
beta = df.iloc[:, 0].values
n = np.arange(1, len(beta) + 1)

# === 2. Modellfunktion: A + a/log(n) + b*cos(...) + c*n ===
def beta_model(n, A, a, b, f, phi, c):
    return A + a / np.log(n) + b * np.cos(2 * np.pi * f * n + phi) + c * n

# === 3. Fit durchführen ===
p0 = [0, 1e5, 1e4, 0.00001, 0, 0]
params, _ = curve_fit(beta_model, n, beta, p0=p0, maxfev=10000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params
beta_fit = beta_model(n, *params)
residual = beta - beta_fit

# === 4. Visualisierung ===
fig, axs = plt.subplots(2, 2, figsize=(16, 8))

axs[0, 0].plot(n, beta, color='gray', label='Reale Beta(n)-Werte')
axs[0, 0].plot(n, beta_fit, color='blue', label='Modellfunktion')
axs[0, 0].set_title('Vergleich: Modellfunktion vs. Beta(n)-Skala')
axs[0, 0].legend()

axs[0, 1].plot(n, residual, color='purple', label='Residuum')
axs[0, 1].axhline(0, linestyle='--', color='gray')
axs[0, 1].set_title('Residuum (Beta - Modell)')
axs[0, 1].legend()

axs[1, 0].plot(n, a_fit / np.log(n), color='orange', label=r'$\frac{a}{\log n}$')
axs[1, 0].set_title('Log-Term')

axs[1, 1].plot(n, b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit), color='red', label='Oszillation')
axs[1, 1].set_title('Kosinus-Term')

for ax in axs.flat:
    ax.set_xlabel('n')
    ax.set_ylabel(r'$\beta(n)$')
    ax.legend()
    ax.axhline(0, linestyle='--', color='gray')

plt.tight_layout()
plt.show()

# === 5. Parameter anzeigen ===
print("✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten laden ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
beta = pd.read_csv(data_path, header=None).iloc[:, 0].astype(float).values

# === 2. Index n definieren (Start bei 2 wegen log(n)) ===
n = np.arange(2, len(beta) + 2)
beta = beta[1:]  # erster Eintrag entfällt, da n bei 2 beginnt

# === 3. Modellfunktion definieren ===
def beta_model(n, A, a, b, f, phi, c):
    return A + a / np.log(n) + b * np.cos(2 * np.pi * f * n + phi) + c * n

# === 4. Startwerte und Grenzen ===
p0 = [0, 1e5, 1e4, 1e-5, 0, 0]
bounds = (
    [-np.inf, 0, 0, 1e-6, -np.pi, -1e-3],
    [np.inf, 1e6, 1e5, 1e-3, np.pi, 1e-3]
)

# === 5. Fit durchführen ===
params, _ = curve_fit(beta_model, n, beta, p0=p0, bounds=bounds, maxfev=20000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params

# === 6. Modell auswerten ===
beta_fit = beta_model(n, *params)
residuum = beta - beta_fit

# === 7. Plots ===
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Vergleich Modell und Daten
axs[0, 0].plot(n, beta, color='gray', label="Reale Beta(n)-Werte")
axs[0, 0].plot(n, beta_fit, color='blue', label="Modellfunktion")
axs[0, 0].set_title("Vergleich: Modellfunktion vs. Beta(n)-Skala")
axs[0, 0].set_ylabel(r"$\beta(n)$")
axs[0, 0].legend()
axs[0, 0].axhline(0, color='black', lw=0.5, ls='--')

# Residuum
axs[0, 1].plot(n, residuum, color='purple', label="Residuum")
axs[0, 1].set_title("Residuum (Beta - Modell)")
axs[0, 1].set_ylabel(r"$\beta(n)$")
axs[0, 1].legend()
axs[0, 1].axhline(0, color='black', lw=0.5, ls='--')

# Log-Term
axs[1, 0].plot(n, a_fit / np.log(n), color='orange', label=r"$\frac{a}{\log n}$")
axs[1, 0].set_title("Log-Term")
axs[1, 0].set_ylabel(r"$\beta(n)$")
axs[1, 0].legend()
axs[1, 0].axhline(0, color='black', lw=0.5, ls='--')

# Kosinus-Term
axs[1, 1].plot(n, b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit), color='red', label="Oszillation")
axs[1, 1].set_title("Kosinus-Term")
axs[1, 1].set_ylabel(r"$\beta(n)$")
axs[1, 1].legend()
axs[1, 1].axhline(0, color='black', lw=0.5, ls='--')

plt.tight_layout()
plt.show()

# === Ausgabe der Fit-Parameter ===
print("\n✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten laden und bereinigen ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
df = pd.read_csv(data_path, header=None, dtype=str)
df = df[df[0].str.replace(',', '.').str.match(r'^-?\d+(\.\d+)?$')]
beta = df[0].astype(float).values

# === 2. Index n definieren (Start bei 2 wegen log(n)) ===
n = np.arange(1, len(beta) + 1)

# === 3. Modellfunktion definieren ===
def beta_model(n, A, a, b, f, phi, c):
    with np.errstate(divide='ignore'):
        log_term = np.where(n > 1, a / np.log(n), 0)
    osc_term = b * np.cos(2 * np.pi * f * n + phi)
    drift_term = c * n
    return A + log_term + osc_term + drift_term

# === 4. Fit durchführen ===
p0 = [0, 1e5, 1e4, 1e-5, 0, 0]  # Startwerte für A, a, b, f, phi, c
params, _ = curve_fit(beta_model, n, beta, p0=p0, maxfev=10000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params

# === 5. Modell berechnen ===
beta_fit = beta_model(n, *params)
residuum = beta - beta_fit
log_term = a_fit / np.log(n)
osc_term = b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit)
drift_term = c_fit * n

# === 6. Plotten ===
fig, axs = plt.subplots(2, 2, figsize=(15, 8))

# Vergleich Original vs Modell
axs[0, 0].plot(n, beta, color='gray', linewidth=1, label="Reale Beta(n)-Werte")
axs[0, 0].plot(n, beta_fit, color='blue', linewidth=1, label="Modellfunktion")
axs[0, 0].set_title("Vergleich: Modellfunktion vs. Beta(n)-Skala")
axs[0, 0].legend()
axs[0, 0].set_ylabel(r"$\beta(n)$")

# Residuum
axs[0, 1].plot(n, residuum, color='purple', linewidth=1, label="Residuum")
axs[0, 1].set_title("Residuum (Beta - Modell)")
axs[0, 1].legend()

# Log-Term
axs[1, 0].plot(n, log_term, color='orange', label=r'$\frac{a}{\log n}$')
axs[1, 0].set_title("Log-Term")
axs[1, 0].legend()

# Kosinus-Term
axs[1, 1].plot(n, osc_term, color='red', label="Oszillation")
axs[1, 1].set_title("Kosinus-Term")
axs[1, 1].legend()

plt.tight_layout()
plt.show()

# === 7. Ausgabe der Parameter ===
print("✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.special import gamma
from scipy.fft import fft, fftfreq, fftshift

# === Modellfunktion: analytische Mellin-Transformation ===
def mellin_model(s, alpha, omega):
    A = gamma(s)
    B = (alpha**2 + omega**2)**(-s / 2)
    C = np.cos(s * np.arctan(omega / alpha))
    return np.abs(A * B * C)

# === Beispiel: künstlich erzeugte Theta(N)-Kurve ===
N = np.logspace(0.5, 2.5, 200)
theta = 0.4 * np.exp(-0.05 * N) * np.cos(2 * np.pi * 0.01 * N)

# === Log-Skala vorbereiten ===
u = np.log(N)
u_uniform = np.linspace(u.min(), u.max(), 512)
theta_uniform = np.interp(u_uniform, u, theta)

# === "Mellin-Transformation" über FFT im log-Bereich ===
mellin_fft = fft(theta_uniform)
freqs = fftfreq(len(u_uniform), d=(u_uniform[1] - u_uniform[0]))
s_vals = fftshift(freqs)
mellin_vals = fftshift(np.abs(mellin_fft))

# === Fit im positiven Frequenzbereich ===
mask = s_vals > 0
s_fit = s_vals[mask][:100]
mellin_fit = mellin_vals[mask][:100]

# === Fit durchführen ===
popt, _ = curve_fit(mellin_model, s_fit, mellin_fit, p0=[0.1, 0.1], maxfev=10000)
alpha_fit, omega_fit = popt

# === Plot ===
plt.figure(figsize=(10, 5))
plt.plot(s_fit, mellin_fit, label="Numerisch (FFT)", lw=2)
plt.plot(s_fit, mellin_model(s_fit, *popt), 'r--', label="Fit: analytisch")
plt.xlabel("Im(s) – Frequenz (log-Bereich)")
plt.ylabel("|Mellin[θ](s)|")
plt.title("Numerisch vs. analytische Mellin-Transformation")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Ergebnis anzeigen
print(f"✅ Gefittete Parameter:\n  α = {alpha_fit:.6f}\n  ω = {omega_fit:.6f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit

# === 1. Daten laden ===
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"  # Pfad ggf. anpassen
beta = pd.read_csv(beta_path, skiprows=1, header=None).squeeze().astype(float).values

# === 2. Logarithmische Achse erzeugen ===
n = np.arange(1, len(beta) + 1)
log_n = np.log(n)

# === 3. Interpolation auf gleichmäßigem Gitter ===
u = np.linspace(log_n.min(), log_n.max(), len(beta))
interp = interp1d(log_n, beta, kind='linear', fill_value="extrapolate")
beta_interp = interp(u)

# === 4. FFT auf logarithmischem Gitter (entspricht numerischer Mellin-Trafo) ===
beta_fft = fft(beta_interp)
freqs = fftfreq(len(u), d=(u[1] - u[0]))  # imaginäre Frequenzen im s-Raum

# === 5. Nur positive Frequenzen betrachten ===
mask = freqs > 0
freqs_pos = freqs[mask]
spectrum = np.abs(beta_fft[mask])

# === 6. Modellfunktion für analytische Mellin-Trafo (z. B. Gammamoduliert) ===
def mellin_analytic(w, alpha, omega):
    return np.cos(w * np.arctan(omega / alpha)) * np.exp(-w * np.arctan(omega / alpha)) / (alpha**2 + omega**2)**(w / 2)

# === 7. Fit auf numerisches Spektrum ===
p0 = [5, 5]
params, _ = curve_fit(mellin_analytic, freqs_pos, spectrum / np.max(spectrum), p0=p0)
alpha_fit, omega_fit = params

# === 8. Plot ===
plt.figure(figsize=(12, 5))
plt.plot(freqs_pos, spectrum / np.max(spectrum), label='Numerisch (FFT)')
plt.plot(freqs_pos, mellin_analytic(freqs_pos, *params), 'r--', label='Fit: analytisch')
plt.xlabel('Im(s) - Frequenz (log-Bereich)')
plt.ylabel(r'$|\mathrm{Mellin}[\beta](s)|$')
plt.title('Numerisch vs. analytische Mellin-Transformation')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Ergebnisse anzeigen ===
print("Gefittete Parameter:")
print(f"  α = {alpha_fit:.6f}")
print(f"  ω = {omega_fit:.6f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq, fftshift
from scipy.interpolate import interp1d
from scipy.optimize import curve_fit
from scipy.special import gamma

# === 1. Daten laden ===
beta_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"  # Pfad ggf. anpassen
beta = pd.read_csv(beta_path, skiprows=1, header=None).squeeze().astype(float).values

# === 2. Logarithmische Achse erzeugen ===
n = np.arange(1, len(beta) + 1)
log_n = np.log(n)

# === 3. Interpolation auf gleichmäßigem Gitter ===
u = np.linspace(log_n.min(), log_n.max(), len(beta))
interp = interp1d(log_n, beta, kind='linear', fill_value="extrapolate")
beta_interp = interp(u)

# === 4. FFT auf logarithmischem Gitter (entspricht numerischer Mellin-Trafo) ===
beta_fft = fft(beta_interp)
freqs = fftfreq(len(u), d=(u[1] - u[0]))  # imaginäre Frequenzen im s-Raum
freqs = fftshift(freqs)
spectrum = np.abs(fftshift(beta_fft))

# === 5. Nur positive Frequenzen betrachten und Bereich einschränken ===
mask = (freqs > 0) & (freqs < 10)  # Frequenzbereich auf [0, 10] beschränken
freqs_pos = freqs[mask]
spectrum_pos = spectrum[mask]

# === 6. Modellfunktion für analytische Mellin-Trafo ===
def mellin_model(w, alpha, omega):
    angle = np.arctan(omega / alpha)
    decay = np.exp(-w * angle)
    amp = (alpha**2 + omega**2)**(-w / 2)
    return np.abs(gamma(w) * np.cos(w * angle) * decay * amp)

# === 7. Fit auf logarithmiertes Spektrum durchführen ===
log_spectrum = np.log(spectrum_pos)
p0 = [1.0, 1.0]  # Startwerte für alpha und omega
params, _ = curve_fit(mellin_model, freqs_pos, np.exp(log_spectrum), p0=p0, maxfev=10000)
alpha_fit, omega_fit = params

# === 8. Plot ===
plt.figure(figsize=(12, 6))
plt.plot(freqs_pos, spectrum_pos, label='Numerisch (FFT)', lw=2)
plt.plot(freqs_pos, mellin_model(freqs_pos, *params), 'r--', label='Fit: analytisch')
plt.xlabel('Im(s) – Frequenz (log-Bereich)')
plt.ylabel(r'$|\mathrm{Mellin}[\beta](s)|$')
plt.title('Numerisch vs. analytische Mellin-Transformation')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Ergebnisse anzeigen ===
print(f"Gefittete Parameter:\n  α = {alpha_fit:.6f}\n  ω = {omega_fit:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Beispieldaten simulieren (hier ersetzen durch echte FFT-Daten) ===
# x_data = Frequenzen (z. B. imaginäre Teile von s)
# y_data = |Mellin[f](s)| aus numerischer Berechnung

x_data = np.linspace(1, 10, 100)
y_data = 1e9 * x_data**-2.3 + np.random.normal(0, 1e6, size=len(x_data))  # Beispiel mit Rauschen

# === 2. Modellfunktion: Potenzgesetz ===
def power_law(x, A, k):
    return A * x**(-k)

# === 3. Fit durchführen ===
p0 = [1e9, 2]  # Startwerte für A und k
params, _ = curve_fit(power_law, x_data, y_data, p0=p0)
A_fit, k_fit = params

# === 4. Visualisierung ===
plt.figure(figsize=(10, 5))
plt.scatter(x_data, y_data, label='Numerisch (FFT)', color='blue', s=10)
plt.plot(x_data, power_law(x_data, A_fit, k_fit), color='red',
         label=f'Fit: A = {A_fit:.2e},  k = {k_fit:.3f}', linewidth=2, linestyle='--')
plt.xlabel('Im(s) – Frequenz (log-Bereich)')
plt.ylabel('|Mellin[f](s)|')
plt.title('Numerisch vs. Fit: Potenzgesetz')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# === 5. Ergebnisse ausgeben ===
print(f"✅ Gefittete Parameter:\n  A = {A_fit:.6e}\n  k = {k_fit:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpmath import zetazero
from scipy.optimize import curve_fit

# === Schritt 1: Reale Zeta-Nullstellen laden
N = 100  # Anzahl der Nullstellen
zeta_zeros = np.array([float(zetazero(i).imag) for i in range(1, N + 1)])

# === Schritt 2: Deine ZNF-Formel
def znf_model(n, A, beta, C):
return A * n**beta + C * np.log(n)

# === Schritt 3: n-Werte
n_vals = np.arange(1, N + 1)

# === Schritt 4: Fit deiner Formel auf echte Zeta-Nullstellen
popt, _ = curve_fit(znf_model, n_vals, zeta_zeros, p0=[2.0, 0.9, 0.0])
A_fit, beta_fit, C_fit = popt

print(f"✅ Angepasste ZNF-Parameter:")
print(f"A = {A_fit:.6f},  β = {beta_fit:.6f},  C = {C_fit:.6f}")

# === Schritt 5: Rekonstruierte Nullstellen
znf_fit = znf_model(n_vals, A_fit, beta_fit, C_fit)

# === Schritt 6: Plot Vergleich ZNF vs. Zeta
plt.figure(figsize=(8, 5))
plt.plot(n_vals, zeta_zeros, label="Zeta-Nullstellen", linewidth=2)
plt.plot(n_vals, znf_fit, '--', label="ZNF-Modell (Fit)", linewidth=2)
plt.xlabel("Index n")
plt.ylabel("Imaginärteil ρ")
plt.title("Rekonstruktion der RH-Nullstellen aus ZNF-Formel")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# === Schritt 7: Abweichung analysieren
delta = znf_fit - zeta_zeros

plt.figure(figsize=(8, 3))
plt.plot(n_vals, delta, color='purple')
plt.axhline(0, linestyle='--', color='gray')
plt.xlabel("Index n")
plt.ylabel("Fehler (ZNF - Zeta)")
plt.title("Abweichung der rekonstruierten Nullstellen")
plt.grid()
plt.tight_layout()
plt.show()

# === Schritt 8: Korrelation & Fehler
correlation = np.corrcoef(zeta_zeros, znf_fit)[0, 1]
mse = np.mean((delta)**2)

print(f"📊 Pearson-Korrelation: {correlation:.8f}")
print(f"📉 MSE der Rekonstruktion: {mse:.8f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# === 1. Daten laden und bereinigen ===
data_path = "/content/drive/MyDrive/beta_rekonstruiert_nur_beta_werte.csv"
df = pd.read_csv(data_path, header=None, dtype=str)
df = df[df[0].str.replace(',', '.').str.match(r'^-?\d+(\.\d+)?$')]
beta = df[0].astype(float).values

# === 2. Index n definieren (Start bei 2 wegen log(n)) ===
n = np.arange(1, len(beta) + 1)

# === 3. Modellfunktion definieren ===
def beta_model(n, A, a, b, f, phi, c):
    with np.errstate(divide='ignore'):
        log_term = np.where(n > 1, a / np.log(n), 0)
    osc_term = b * np.cos(2 * np.pi * f * n + phi)
    drift_term = c * n
    return A + log_term + osc_term + drift_term

# === 4. Fit durchführen ===
p0 = [0, 1e5, 1e4, 1e-5, 0, 0]  # Startwerte für A, a, b, f, phi, c
params, _ = curve_fit(beta_model, n, beta, p0=p0, maxfev=10000)
A_fit, a_fit, b_fit, f_fit, phi_fit, c_fit = params

# === 5. Modell berechnen ===
beta_fit = beta_model(n, *params)
residuum = beta - beta_fit
log_term = a_fit / np.log(n)
osc_term = b_fit * np.cos(2 * np.pi * f_fit * n + phi_fit)
drift_term = c_fit * n

# === 6. Plotten ===
fig, axs = plt.subplots(2, 2, figsize=(15, 8))

# Vergleich Original vs Modell
axs[0, 0].plot(n, beta, color='gray', linewidth=1, label="Reale Beta(n)-Werte")
axs[0, 0].plot(n, beta_fit, color='blue', linewidth=1, label="Modellfunktion")
axs[0, 0].set_title("Vergleich: Modellfunktion vs. Beta(n)-Skala")
axs[0, 0].legend()
axs[0, 0].set_ylabel(r"$\beta(n)$")

# Residuum
axs[0, 1].plot(n, residuum, color='purple', linewidth=1, label="Residuum")
axs[0, 1].set_title("Residuum (Beta - Modell)")
axs[0, 1].legend()

# Log-Term
axs[1, 0].plot(n, log_term, color='orange', label=r'$\frac{a}{\log n}$')
axs[1, 0].set_title("Log-Term")
axs[1, 0].legend()

# Kosinus-Term
axs[1, 1].plot(n, osc_term, color='red', label="Oszillation")
axs[1, 1].set_title("Kosinus-Term")
axs[1, 1].legend()

plt.tight_layout()
plt.show()

# === 7. Ausgabe der Parameter ===
print("✅ Gefittete Parameter:")
print(f"  A   = {A_fit:.4f}")
print(f"  a   = {a_fit:.4f}")
print(f"  b   = {b_fit:.4f}")
print(f"  f   = {f_fit:.6f}")
print(f"  φ   = {phi_fit:.4f}")
print(f"  c   = {c_fit:.6e}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Dateipfade
pfad_nullstellen = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"
pfad_beta = "/content/drive/MyDrive/beta_scale_operator_cleaned.csv"

# Nullstellen einlesen
nullstellen = pd.read_csv(pfad_nullstellen, header=None).squeeze("columns").astype(float).values
nullstellen_diffs = np.diff(nullstellen)
nullstellen_diffs_norm = (nullstellen_diffs - np.mean(nullstellen_diffs)) / np.std(nullstellen_diffs)

# Beta-Werte einlesen und spiegeln
beta = pd.read_csv(pfad_beta, header=None).squeeze("columns").astype(float).values
beta = beta[::-1]
beta_diffs = np.diff(beta)
beta_diffs_norm = (beta_diffs - np.mean(beta_diffs)) / np.std(beta_diffs)

# FFT-Funktion
def plot_fft(signal, title):
    n = len(signal)
    freq = fftfreq(n, d=1)[:n // 2]
    spectrum = np.abs(fft(signal))[:n // 2]

    plt.figure(figsize=(10, 4))
    plt.plot(freq, spectrum)
    plt.title(f"FFT: {title}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# FFT der normierten Differenzen
plot_fft(nullstellen_diffs_norm, "Nullstellen-Differenzen (normiert)")
plot_fft(-beta_diffs_norm, "Gespiegelte Beta-Differenzen (normiert)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
from scipy.stats import pearsonr

# === Datei-Pfade ===
pfad_beta = "/content/drive/MyDrive/beta_scale_operator_cleaned.csv"
pfad_nullstellen = "/content/drive/MyDrive/Aus Chrome gespeichert/zeros6_fixed.csv"

# === Daten einlesen ===

# Beta-Werte laden
beta_df = pd.read_csv(pfad_beta, header=None)
beta_werte = beta_df[0].astype(float).values

# Nullstellen (nur Imaginärteil vorhanden)
nullstellen_df = pd.read_csv(pfad_nullstellen, header=None)
imag_parts = nullstellen_df.iloc[:, 0].astype(float).values

# === Differenzen berechnen ===
beta_diffs = np.diff(beta_werte)[::-1]  # gespiegelte Differenzen
nullstellen_diffs = np.diff(imag_parts)

# === Kürzen auf gleiche Länge ===
min_len = min(len(beta_diffs), len(nullstellen_diffs))
beta_diffs = beta_diffs[:min_len]
nullstellen_diffs = nullstellen_diffs[:min_len]

# === Normalisieren ===
beta_diffs_norm = (beta_diffs - np.mean(beta_diffs)) / np.std(beta_diffs)
nullstellen_diffs_norm = (nullstellen_diffs - np.mean(nullstellen_diffs)) / np.std(nullstellen_diffs)

# === Pearson-Korrelation ===
corr, _ = pearsonr(nullstellen_diffs_norm, beta_diffs_norm)
print(f"Pearson-Korrelation (Nullstellen-Differenzen vs. gespiegelte Beta-Differenzen): {corr:.4f}")

# === Vergleichsplot ===
plt.figure(figsize=(14, 5))
plt.plot(nullstellen_diffs_norm, label="Normierte Nullstellen-Differenzen")
plt.plot(beta_diffs_norm, label="Normierte, gespiegelte Beta-Differenzen", linestyle="--")
plt.title("Vergleich: Nullstellen vs. gespiegelte Beta-Skala (normiert)")
plt.xlabel("Index")
plt.ylabel("Normierte Differenz")
plt.legend()
plt.tight_layout()
plt.show()

# === FFT-Funktion ===
def plot_fft(signal, label, n=10):
    spectrum = np.abs(fft(signal))
    spectrum = spectrum[:len(spectrum)//2]
    freqs = np.fft.fftfreq(len(signal), d=1)[:len(spectrum)]

    # Visualisierung
    plt.figure(figsize=(12, 4))
    plt.plot(freqs, spectrum)
    plt.title(f"FFT: {label}")
    plt.xlabel("Frequenz")
    plt.ylabel("Amplitude")
    plt.tight_layout()
    plt.show()

    # Dominante Frequenzen ausgeben
    top_indices = np.argsort(spectrum)[-n:][::-1]
    print(f"Top {n} Frequenzen ({label}):")
    for i in top_indices:
        print(f"  Frequenz: {freqs[i]:.4f}, Stärke: {spectrum[i]:.2f}")

# === FFT-Analyse ===
plot_fft(nullstellen_diffs_norm, "Nullstellen-Differenzen (normiert)")
plot_fft(beta_diffs_norm, "Gespiegelte Beta-Differenzen (normiert)")
```
